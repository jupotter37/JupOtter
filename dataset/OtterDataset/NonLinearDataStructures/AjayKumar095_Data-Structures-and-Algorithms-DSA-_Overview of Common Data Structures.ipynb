<h1 style='color: green;'><center>Overview of Common Data Structures</center></h1>

<h3>In the world of DSA, there are many types of Data Structure that we use to solve the data science problems. Following are some most common data structures that we use.</h3>

<h4 style='color: red;'>Arrays</h4>
<p><strong>Definition:</strong> A collection of elements, each identified by at least one array index or key. All elements are of same type.</p>

<p><strong>Properties of Array Data Structure:</strong></p>
<ul>
<li><b>Fixed Size:</b> The size of the array is defined at the time of creation and cannot be changed.</li>
<li><b>Contiguous Memory Location:</b> Elements are stored in contiguous memory locatio, which allows constant-time access of elements.</li>
<li><b>Indexing:</b> All the elements of array is represented by its index value, and in array the index start from 0, that means the first element in at index 0.</li>
</ul>

<p><strong>Operations:</strong></p>
<ul>
<li><b>Access:</b> big O(1) time complexity.</li>
<li><b>Search:</b> big O(n) time complexity in the worst case (linear search).</li>
<li><b>Insertion/Deletion:</b> big O(n) time complexity as elements need to be shifted.</li>
</ul>

<p><strong>Use Cases of Array:</strong></p>
<ul>
<li>Efficient random access to elements.</li>
<li>suitable for static data.</li>
<li>Example: Implementing a lookup table.</li>
</ul>

<h4 style='color: red;'>Linked Lists</h4>
<p><b>Definition: </b>A linear data structure where elemesnts, called nodes, are linked using pointers. Each node of linked list divide into two parts one store the data and other part store the location or pointer of next node.</p>

<p><b>Type of linked list:</b></p>
<ul>
<li><b>Singly Linked List: </b>Each node contains data and a pointer to the next node.</li>
<li><b>Dubly Linked List: </b>Each node contains data, a pointer to the next node, and a pointer to the previous node.</li>
<li><b>Circular Linked List: </b>The last node points to the first node, forming a circle.</li>
</ul>

<p><b>Properties of linked list:</b></p>
<ul>
<li><b>Dynamic Size: </b>The list grows and shrinks as needed.</li>
<li><b>Memory Utilization: </b>Node are not stored in contiguous memory locations.</li>
</ul>

<p><b>Operations:</b></p>
<ul>
<li><b>Access: </b>big O(n) time complexith to access an element.</li>
<li><b>Insertion/Deletion: </b>big O(1) time complexity if the node to insert is known.</li>
<li><b>Search: </b>big O(n) time complexity.</li>
</ul>

<p><b>Use Cases:</b></p>
<ul>
<li>Efficient insertions and deletions.</li>
<li>Implementation of other data structures (stacks, queues.)</li>
<li>Example: Navigation system where you need to go forward and backward.</li>
</ul>

<h4 style='color: red;'>Stacks</h4>
<p><b>Definition: </b>A collection of elements with two main operations: push (to add an element to the top) and pop (remove the top element.). It like a container that is closed by it bottom and their is only one way to add or remove the data from the container. The last element remove first and the first element is remove at the last. It work on LIFO priciple means Last In First Out.</p>

<p><b>Properties of Stacks</b></p>
<ul>
   <li><b>LIFO Order: </b>Last In, First Out.</li>
   <li><b>Operations: </b>Push, pop, and peek.</li>
</ul>

<p><b>Operations:</b></p>
<ul>
    <li>Push: O(1) time complexity.</li>
    <li>Pop: O(1) time complexity.</li>
    <li>Peek: O(1) time complexity.</li>
</ul>
<p><b>Use Cases:</b></p>
<ul>
     <li>Backtracking algorithms (e.g., solving mazes).</li>
    <li>Function call management (call stack).</li>
     <li>Example: Undo feature in text editors.</li>
 </ul>
 


<h4 style='color: red;'>Queues</h4>
<p><strong>Definition:</strong> A collection of elements that follows the FIFO principle, where elements are added at the rear and removed from the front. It like a hollow pipe in which the water flow, the water insert from one end and come out form the other side of the pipe.</p>
<ul>
        <li><strong>Properties:</strong>
            <ul>
                <li>FIFO Order: First In, First Out.</li>
                <li>Operations: Enqueue (add an element), dequeue (remove an element), and front (view the front element).</li>
            </ul>
        </li>
        <li><strong>Operations:</strong>
            <ul>
                <li>Enqueue: O(1) time complexity.</li>
                <li>Dequeue: O(1) time complexity.</li>
                <li>Front: O(1) time complexity.</li>
            </ul>
        </li>
        <li><strong>Use Cases:</strong>
            <ul>
                <li>Task scheduling in operating systems.</li>
                <li>Handling asynchronous data (e.g., IO buffers).</li>
                <li>Example: Print queues in printers.</li>
            </ul>
        </li>
    </ul>

<h4 style='color: red;'>Tree</h4>
<p><strong>Definition:</strong> A hierarchical data structure consisting of nodes, with a single root node and child nodes forming a parent-child relationship.</p>
    <ul>
        <li><strong>Types:</strong>
            <ul>
                <li>Binary Tree: Each node has at most two children.</li>
                <li>Binary Search Tree (BST): A binary tree with ordered nodes, left child is smaller, and right child is larger.</li>
                <li>AVL Tree: A self-balancing BST where the height difference between left and right subtrees is at most one.</li>
                <li>Red-Black Tree: A self-balancing BST with additional properties ensuring balance.</li>
                <li>Heap: A special tree where the parent node is greater (max-heap) or smaller (min-heap) than its children.</li>
            </ul>
        </li>
        <li><strong>Properties:</strong>
            <ul>
                <li>Hierarchical Structure: Represents hierarchical relationships.</li>
                <li>Balanced Trees: Ensure O(log n) operations.</li>
            </ul>
        </li>
        <li><strong>Operations:</strong>
            <ul>
                <li>Insertion: O(log n) in balanced trees.</li>
                <li>Deletion: O(log n) in balanced trees.</li>
                <li>Traversal: O(n), common types include in-order, pre-order, and post-order.</li>
            </ul>
        </li>
        <li><strong>Use Cases:</strong>
            <ul>
                <li>Hierarchical data representation (e.g., file systems).</li>
                <li>Efficient searching and sorting (BST).</li>
                <li>Example: Managing hierarchical data like organization structures.</li>
            </ul>
        </li>
    </ul>

<h4 style='color: red;'>Graph</h4>
 <p><strong>Definition:</strong> A collection of nodes (vertices) and edges connecting pairs of nodes, used to represent relationships.</p>
    <ul>
        <li><strong>Types:</strong>
            <ul>
                <li>Directed Graph: Edges have directions.</li>
                <li>Undirected Graph: Edges have no directions.</li>
                <li>Weighted Graph: Edges have weights.</li>
            </ul>
        </li>
        <li><strong>Properties:</strong>
            <ul>
                <li>Vertices and Edges: Basic components.</li>
                <li>Adjacency Matrix/List: Common representations.</li>
            </ul>
        </li>
        <li><strong>Operations:</strong>
            <ul>
                <li>Traversal: DFS (Depth First Search) and BFS (Breadth First Search).</li>
                <li>Shortest Path: Algorithms like Dijkstraâ€™s and Bellman-Ford.</li>
            </ul>
        </li>
        <li><strong>Use Cases:</strong>
            <ul>
                <li>Network topology (computer networks, social networks).</li>
                <li>Pathfinding (e.g., GPS).</li>
                <li>Example: Social network relationships.</li>
            </ul>
        </li>
    </ul>

<h4 style='color: red;'>Hash Tables</h4>
<p><strong>Definition:</strong> A data structure that maps keys to values using a hash function to compute an index.</p>
    <ul>
        <li><strong>Properties:</strong>
            <ul>
                <li>Hash Function: Computes an index for each key.</li>
                <li>Collision Handling: Techniques like chaining or open addressing.</li>
            </ul>
        </li>
        <li><strong>Operations:</strong>
            <ul>
                <li>Insertion: O(1) average time complexity.</li>
                <li>Deletion: O(1) average time complexity.</li>
                <li>Search: O(1) average time complexity.</li>
            </ul>
        </li>
        <li><strong>Use Cases:</strong>
            <ul>
                <li>Implementing associative arrays (dictionaries).</li>
                <li>Efficient data retrieval.</li>
                <li>Example: Caching.</li>
            </ul>
        </li>
    </ul>

<h4 style='color: red;'>Heaps</h4>
<p><strong>Definition:</strong> A special tree-based data structure that satisfies the heap property: max-heap (parent is greater than children) or min-heap (parent is smaller than children).</p>
    <ul>
        <li><strong>Properties:</strong>
            <ul>
                <li>Heap Property: Ensures efficient access to the maximum or minimum element.</li>
                <li>Complete Binary Tree: Heaps are usually implemented as complete binary trees.</li>
            </ul>
        </li>
        <li><strong>Operations:</strong>
            <ul>
                <li>Insertion: O(log n) time complexity.</li>
                <li>Deletion (extract-max/min): O(log n) time complexity.</li>
                <li>Peek: O(1) time complexity.</li>
            </ul>
        </li>
        <li><strong>Use Cases:</strong>
            <ul>
                <li>Implementing priority queues.</li>
                <li>Efficiently finding the minimum or maximum element.</li>
                <li>Example: Task scheduling in operating systems.</li>
            </ul>
        </li>
    </ul>
