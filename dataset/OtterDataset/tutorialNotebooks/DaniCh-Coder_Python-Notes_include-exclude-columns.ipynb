```python
import pandas as pd
```

This tutorial uses the Titanic data set, stored as CSV. The data consists of the following data columns:

PassengerId: Id of every passenger.

Survived: Indication whether passenger survived. 0 for yes and 1 for no.

Pclass: One out of the 3 ticket classes: Class 1, Class 2 and Class 3.

Name: Name of passenger.

Sex: Gender of passenger.

Age: Age of passenger in years.

SibSp: Number of siblings or spouses aboard.

Parch: Number of parents or children aboard.

Ticket: Ticket number of passenger.

Fare: Indicating the fare.

Cabin: Cabin number of passenger.

Embarked: Port of embarkation.

## Selecting Columns by type - Include - Exclude

```python
# include all the columns type object
titanic.select_dtypes(include='object')
```

```python
# exclude columns of type object
titanic.select_dtypes(exclude='object')
```

```python
# select int64 type columns
titanic.select_dtypes(include='int')
```

Include and exclude of the same type on the same dataframe are complementaries.
* the sum of the columns are equal to total columns

```python
# select float columns
titanic.select_dtypes(include='float')
```

```python
# select category columns - there is no category columns
titanic.select_dtypes(include='category')
```

```python
# select datetime columns - be aware there is no such a column like this
titanic.select_dtypes(include='datetime')
```

## Drop columns

### Drop Columns by type

```python
# first select a dataframe of a dtype
column_type_to_delete = titanic.select_dtypes(include=['float64','int64'])

# then get a list with column names
list_columns_selected = list(column_type_to_delete)

# finally drop the columns of the dataframe
titanic.drop(list_columns_selected, axis=1, inplace = True) # add implace = True to delete definitively
```
