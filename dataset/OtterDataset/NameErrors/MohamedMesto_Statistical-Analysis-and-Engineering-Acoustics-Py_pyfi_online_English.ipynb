   * Write a function that receives a vector and alternately adds or subtracts its squared entries. *(3 points)*
   * The calculation for a vector $\pmb{b}$ is:
$$
{b_1}^2 - {b_2}^2 + {b_3}^2 - {b_4}^2 + {b_5}^2 \ldots
$$

   * The function should return the scalar result of the calculation. * (1 point) *
   * Use NumPy arrays and avoid if clauses and loops if possible. * (1 point) *
   * Apply the function to the following vector and output the result of the calculation: * (1 point) *

$~~~~~\pmb{b}=~\big($ `3.2, -2.7, 0.2, 3.8, -1.9, -2.1, -3.5, 1.4, -1.8, 2.7, -2.1, -1.6, -0.3, -1.0, -1.2` $\big)$
*********

```python
# Hier eigenen Code schreiben ...

###Answer:3
print('***### Answer3: ###***\n')

def calculate(b):
    b=np.array(b)
    even_summe = np.sum(b[0:len(b):2]**2)
    odd_summe = np.sum(b[1:len(b):2]**2)
    result = even_summe-odd_summe
    return result
    
b=( 3.2, -2.7, 0.2, 3.8, -1.9, -2.1, -3.5, 1.4, -1.8, 2.7, -2.1, -1.6, -0.3, -1.0, -1.2 )

print(calculate(b))
print('\n**************************************\n')
```
