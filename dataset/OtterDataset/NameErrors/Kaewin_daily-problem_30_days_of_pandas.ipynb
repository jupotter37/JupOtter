# 1 Big Countries

Table: ```World```

```
+-------------+---------+
| Column Name | Type    |
+-------------+---------+
| name        | varchar |
| continent   | varchar |
| area        | int     |
| population  | int     |
| gdp         | bigint  |
+-------------+---------+
``````
name is the primary key (column with unique values) for this table.
Each row of this table gives information about the name of a country, the continent to which it belongs, its area, the population, and its GDP value.

 

A country is big if:

    it has an area of at least three million (i.e., 3000000 km2), or
    it has a population of at least twenty-five million (i.e., 25000000).

Write a solution to find the name, population, and area of the big countries.

Return the result table in any order.

The result format is in the following example.

 

Example 1:
``````
Input: 
World table:
+-------------+-----------+---------+------------+--------------+
| name        | continent | area    | population | gdp          |
+-------------+-----------+---------+------------+--------------+
| Afghanistan | Asia      | 652230  | 25500100   | 20343000000  |
| Albania     | Europe    | 28748   | 2831741    | 12960000000  |
| Algeria     | Africa    | 2381741 | 37100000   | 188681000000 |
| Andorra     | Europe    | 468     | 78115      | 3712000000   |
| Angola      | Africa    | 1246700 | 20609294   | 100990000000 |
+-------------+-----------+---------+------------+--------------+
Output: 
+-------------+------------+---------+
| name        | population | area    |
+-------------+------------+---------+
| Afghanistan | 25500100   | 652230  |
| Algeria     | 37100000   | 2381741 |
+-------------+------------+---------+
``````

```````python
import pandas as pd
data = [['Afghanistan', 'Asia', 652230, 25500100, 20343000000], ['Albania', 'Europe', 28748, 2831741, 12960000000], ['Algeria', 'Africa', 2381741, 37100000, 188681000000], ['Andorra', 'Europe', 468, 78115, 3712000000], ['Angola', 'Africa', 1246700, 20609294, 100990000000]]
World = pd.DataFrame(data, columns=['name', 'continent', 'area', 'population', 'gdp']).astype({'name':'object', 'continent':'object', 'area':'Int64', 'population':'Int64', 'gdp':'Int64'})
```````

```````python
import pandas as pd

def big_countries( world: pd.DataFrame) -> pd.DataFrame:

    """A country is big if:

    it has an area of at least three million (i.e., 3000000 km2), or
    it has a population of at least twenty-five million (i.e., 25000000).

    Write a solution to find the name, population, and area of the big countries.

    Return the result table in any order."""

    big_countries_df = world[(world['area'] >= 3000000) | (world['population'] >= 25000000)]

    big_countries_df.drop(columns=['continent', 'gdp'], inplace=True)

    return big_countries_df

big_countries(World)
```````

Write a solution to find the ids of products that are both low fat and recyclable.


```````python
import pandas as pd

data = [[1, 'Joe'], [2, 'Henry'], [3, 'Sam'], [4, 'Max']]
Customers = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})
data = [[1, 3], [2, 1]]
Orders = pd.DataFrame(data, columns=['id', 'customerId']).astype({'id':'Int64', 'customerId':'Int64'})
```````

```````python
import pandas as pd

def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:

    result = pd.merge(orders, customers, how='right', left_on='customerId', right_on='id')
    result.drop(columns=["id_x", "id_y"], inplace=True)
    # drop rows where customerId has a value
    result = result[result['customerId'].isna()]
    result.drop(columns=["customerId"], inplace=True)
    result.rename(columns={'name': 'Customers'}, inplace=True)  # Rename 'id' column from orders

    return result

find_customers(Customers, Orders)
```````

```````python
import pandas as pd

data = [[1, 3, 5, '2019-08-01'], [1, 3, 6, '2019-08-02'], [2, 7, 7, '2019-08-01'], [2, 7, 6, '2019-08-02'], [4, 7, 1, '2019-07-22'], [3, 4, 4, '2019-07-21'], [3, 4, 4, '2019-07-21']]
Views = pd.DataFrame(data, columns=['article_id', 'author_id', 'viewer_id', 'view_date']).astype({'article_id':'Int64', 'author_id':'Int64', 'viewer_id':'Int64', 'view_date':'datetime64[ns]'})
```````

```````python
Views
```````

```````python
import pandas as pd

def article_views(views: pd.DataFrame) -> pd.DataFrame:

    df = views[(views['author_id'] == views['viewer_id'])]
    # remove duplicates
    df.drop_duplicates(subset=['viewer_id', 'author_id'], inplace=True)
    df.drop(columns=['article_id', 'author_id', 'view_date'], inplace=True)
    df.rename(columns={'viewer_id': 'id'}, inplace=True)
    df.sort_values(by='id', ascending=True, inplace=True)

    return df

article_views(Views)
```````

# 6 calculate special bonus

Table: Employees
``````
+-------------+---------+
| Column Name | Type    |
+-------------+---------+
| employee_id | int     |
| name        | varchar |
| salary      | int     |
+-------------+---------+
employee_id is the primary key (column with unique values) for this table.
Each row of this table indicates the employee ID, employee name, and salary.
``````
 

Write a solution to calculate the bonus of each employee. The bonus of an employee is 100% of their salary if the ID of the employee is an odd number and the employee's name does not start with the character 'M'. The bonus of an employee is 0 otherwise.

Return the result table ordered by employee_id.

The result format is in the following example.

 

Example 1:
``````
Input: 
Employees table:
+-------------+---------+--------+
| employee_id | name    | salary |
+-------------+---------+--------+
| 2           | Meir    | 3000   |
| 3           | Michael | 3800   |
| 7           | Addilyn | 7400   |
| 8           | Juan    | 6100   |
| 9           | Kannon  | 7700   |
+-------------+---------+--------+
Output: 
+-------------+-------+
| employee_id | bonus |
+-------------+-------+
| 2           | 0     |
| 3           | 0     |
| 7           | 7400  |
| 8           | 0     |
| 9           | 7700  |
+-------------+-------+
Explanation: 
The employees with IDs 2 and 8 get 0 bonus because they have an even employee_id.
The employee with ID 3 gets 0 bonus because their name starts with 'M'.
The rest of the employees get a 100% bonus.
``````


```````python
import pandas as pd

data = [[2, 'Meir', 3000], [3, 'Michael', 3800], [7, 'Addilyn', 7400], [8, 'Juan', 6100], [9, 'Kannon', 7700]]
Employees = pd.DataFrame(data, columns=['employee_id', 'name', 'salary']).astype({'employee_id':'int64', 'name':'object', 'salary':'int64'})
```````

```````python
import pandas as pd

def calculate_special_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    
    employees['bonus'] = employees.apply(lambda row: row['salary'] if (row['employee_id'] % 2 != 0) and (row['name'][0].lower() != 'm') else 0, axis=1)

    employees = employees[['employee_id', 'bonus']]

    employees = employees.sort_values(by=['employee_id'])

    return employees

calculate_special_bonus(Employees)
```````

```````python
import pandas as pd

def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:

    # Write a solution to find the nth highest salary from the Employee table. If there is no nth highest salary, return null.

    # create a set of unique salaries
    unique_salaries = set(employee['Salary'])

    # sort the unique salaries in descending order
    unique_salaries = sorted(unique_salaries, reverse=True)

    # if N is greater than the number of unique salaries, return null
    if N > len(unique_salaries):
        return None

    # return the nth highest salary
    return unique_salaries[N-1]

nth_highest_salary(employee, 2)
```````

```````python
import pandas as pd

def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:

    # Sort by salary
    employee.sort_values(by='salary', inplace=True, ascending=False)

    if len (employee['salary']) == 1:

        # Create a dictionary with the data
        data = {'SecondHighestSalary': [None]}

        # Create the DataFrame
        df = pd.DataFrame(data)

        return df

    if (len (employee['salary'].unique()) == 1) and (len(employee['salary']) != 1):

        # Create a dictionary with the data
        data = {'SecondHighestSalary': [None]}

        # Create the DataFrame
        df = pd.DataFrame(data)

        return df
    
    else:

        second_highest_salary = employee['salary'].unique()[1]

        # Get the second highest salary employees
        second_highest_salary_employees = employee[employee['salary'] == second_highest_salary]

        # drop the id column
        second_highest_salary_employees.drop('id', axis=1, inplace=True)

        # rename the salary column to 'SecondHighestSalary'
        second_highest_salary_employees.rename(columns={'salary': 'SecondHighestSalary'}, inplace=True)

        # drop any duplicate rows
        second_highest_salary_employees.drop_duplicates(inplace=True)

        return second_highest_salary_employees
```````

Table: Delivery
``````
+-----------------------------+---------+
| Column Name                 | Type    |
+-----------------------------+---------+
| delivery_id                 | int     |
| customer_id                 | int     |
| order_date                  | date    |
| customer_pref_delivery_date | date    |
+-----------------------------+---------+
delivery_id is the primary key (column with unique values) of this table.
The table holds information about food delivery to customers that make orders at some date and specify a preferred delivery date (on the same order date or after it).
``````
 

If the customer's preferred delivery date is the same as the order date, then the order is called immediate; otherwise, it is called scheduled.

Write a solution to find the percentage of immediate orders in the table, rounded to 2 decimal places.

The result format is in the following example.

 

Example 1:
``````
Input: 
Delivery table:
+-------------+-------------+------------+-----------------------------+
| delivery_id | customer_id | order_date | customer_pref_delivery_date |
+-------------+-------------+------------+-----------------------------+
| 1           | 1           | 2019-08-01 | 2019-08-02                  |
| 2           | 5           | 2019-08-02 | 2019-08-02                  |
| 3           | 1           | 2019-08-11 | 2019-08-11                  |
| 4           | 3           | 2019-08-24 | 2019-08-26                  |
| 5           | 4           | 2019-08-21 | 2019-08-22                  |
| 6           | 2           | 2019-08-11 | 2019-08-13                  |
+-------------+-------------+------------+-----------------------------+
Output: 
+----------------------+
| immediate_percentage |
+----------------------+
| 33.33                |
+----------------------+
Explanation: The orders with delivery id 2 and 3 are immediate while the others are scheduled.
``````

```````python
import pandas as pd

data = [[1, 1, '2019-08-01', '2019-08-02'], [2, 5, '2019-08-02', '2019-08-02'], [3, 1, '2019-08-11', '2019-08-11'], [4, 3, '2019-08-24', '2019-08-26'], [5, 4, '2019-08-21', '2019-08-22'], [6, 2, '2019-08-11', '2019-08-13']]
Delivery = pd.DataFrame(data, columns=['delivery_id', 'customer_id', 'order_date', 'customer_pref_delivery_date']).astype({'delivery_id':'Int64', 'customer_id':'Int64', 'order_date':'datetime64[ns]', 'customer_pref_delivery_date':'datetime64[ns]'})
```````

```````python
import pandas as pd

def food_delivery(delivery: pd.DataFrame) -> pd.DataFrame:

    # add column order_status
    # If customer_pref_delivery_date  is the same as order_date , then the order is called immediate; otherwise, it is called scheduled.
    delivery['order_status'] = delivery.apply(lambda row: 'immediate' if row['order_date'] == row['customer_pref_delivery_date'] else 'scheduled', axis=1)

    # find the percentage of immediate orders in the dataframe and add it to a new dataframe with one row and one column named immediate_percentage
    immediate_percentage = delivery[delivery['order_status'] == 'immediate']['order_status'].count() / delivery['order_status'].count()
    immediate_percentage = pd.DataFrame([immediate_percentage], columns=['immediate_percentage'])

    # multiply the percentage by 100 and round it to two decimal places
    immediate_percentage['immediate_percentage'] = immediate_percentage['immediate_percentage'].apply(lambda x: round(x * 100, 2))

    return immediate_percentage

food_delivery(Delivery)
```````

```````python
import pandas as pd

def students_and_examinations(students: pd.DataFrame, subjects: pd.DataFrame, examinations: pd.DataFrame) -> pd.DataFrame:

    # Join the three DataFrames on `student_id` and `subject_name`.
    df = examinations.merge(students, on="student_id").merge(subjects, on="subject_name")

    # Count the number of times each student has taken each exam
    df = df.groupby(["student_id", "subject_name"])["subject_name"].count().reset_index(name="num_exams")

    # if the student did not take an exam, set the value to 0
    df["num_exams"] = df["num_exams"].fillna(0)

    # Sort the DataFrame by `student_id` and `subject_name`.
    df = df.sort_values(["student_id", "subject_name"])

    return df

students_and_examinations(Students, Subjects, Examinations)
```````

# 28 manager with at least 5 direct reports

```````python
import pandas as pd

data = [[101, 'John', 'A', None], [102, 'Dan', 'A', 101], [103, 'James', 'A', 101], [104, 'Amy', 'A', 101], [105, 'Anne', 'A', 101], [106, 'Ron', 'B', 101]]
Employee = pd.DataFrame(data, columns=['id', 'name', 'department', 'managerId']).astype({'id':'Int64', 'name':'object', 'department':'object', 'managerId':'Int64'})
```````

```````python
import pandas as pd

def find_managers(employee: pd.DataFrame) -> pd.DataFrame:

    # Write a solution to find managers with at least five direct reports.

    # get a dataframe with the managerId's and the number of direct reports
    df = employee.groupby(by=['managerId'], sort=False)['id'].count().reset_index(name='cnt')

    # filter out the managers with less than 5 direct reports
    df = df[df['cnt'] >= 5]

    # drop the cnt column
    df.drop(columns=['cnt'], inplace=True)

    # rename the id column to managerId
    df.rename(columns={'id': 'managerId'}, inplace=True)

    # merge with the employee table to get the manager names
    df = pd.merge(df, employee, how='left', left_on='managerId', right_on='id')

    # drop the id column
    df.drop(columns=['id'], inplace=True)

    # drop the managerId column
    df.drop(columns=['managerId_x'], inplace=True)

    # drop the department column
    df.drop(columns=['department'], inplace=True)

    # drop the managerId column
    df.drop(columns=['managerId_y'], inplace=True)

    # if df has nulls drop them
    df.dropna(inplace=True)

    return df

find_managers(Employee)
```````
