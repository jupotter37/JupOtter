This assumes that all images are stored in a single directory and needs to be split, and should produce the same result each time (as long as code is unaltered)

```python
import os

# Get all filenames from dataset

def get_files_in_subdirectories(directory):
    file_names = []
    for root, directories, files in os.walk(directory):
        for file in files:
            file_names.append(os.path.basename(file))
    return file_names

# Original directory containing all images
directory_path = r'C:\Users\andre\dev\BSc-Project\Dataset\all_images'
file_names = get_files_in_subdirectories(directory_path)
print(file_names)
```

```python
import random as rd

# Split names of data into training, validation and test sets

file_names.sort()  # make sure that the filenames have a fixed order before shuffling
rd.seed(230)
rd.shuffle(file_names) # shuffles the ordering of filenames (deterministic given the chosen seed)

split_1 = int(0.8 * len(file_names))
split_2 = int(0.9 * len(file_names))
train_filenames = file_names[:split_1]
val_filenames = file_names[split_1:split_2]
test_filenames = file_names[split_2:]
```

```python
# Define directories for train, validation, and test data
train_dir = r'C:\Users\andre\dev\BSc-Project\Dataset\data\train'
val_dir = r'C:\Users\andre\dev\BSc-Project\Dataset\data\val'
test_dir = r'C:\Users\andre\dev\BSc-Project\Dataset\data\test'
```

```python
import shutil

# Move images from original to dataset to the new datasets according to the split above

# Move images to respective directories
def move_images(filenames, source_dir, destination_dir):
    for filename in filenames:
        source_path = os.path.join(source_dir, filename)
        destination_path = os.path.join(destination_dir, filename)
        shutil.move(source_path, destination_path)

source_directory_path = r'C:\Users\andre\dev\BSc-Project\Dataset\all_images'

move_images(train_filenames, source_directory_path, train_dir)
move_images(val_filenames, source_directory_path, val_dir)
move_images(test_filenames, source_directory_path, test_dir)  
```

```python
# Sort splitted data into classes of different skin lesions:
# ACK, BCC, MEL, NEV, SCC, SEK

import csv

# Path to the CSV file containing metadata
csv_file = r'C:\Users\andre\dev\BSc-Project\Dataset\metadata.csv'


# Create or link paths to the classes within each dataset
base_dir = r'C:\Users\andre\dev\BSc-Project\Dataset\data'

# Create a dictionary to map class names to directories
class_directories_train = {
    "ACK": r'C:\Users\andre\dev\BSc-Project\Dataset\data\train\ACK',
    "BCC": r'C:\Users\andre\dev\BSc-Project\Dataset\data\train\BCC',
    "MEL": r'C:\Users\andre\dev\BSc-Project\Dataset\data\train\MEL',
    "NEV": r'C:\Users\andre\dev\BSc-Project\Dataset\data\train\NEV',
    "SCC": r'C:\Users\andre\dev\BSc-Project\Dataset\data\train\SCC',
    "SEK": r'C:\Users\andre\dev\BSc-Project\Dataset\data\train\SEK',    
}

class_directories_val = {
    "ACK": r'C:\Users\andre\dev\BSc-Project\Dataset\data\val\ACK',
    "BCC": r'C:\Users\andre\dev\BSc-Project\Dataset\data\val\BCC',
    "MEL": r'C:\Users\andre\dev\BSc-Project\Dataset\data\val\MEL',
    "NEV": r'C:\Users\andre\dev\BSc-Project\Dataset\data\val\NEV',
    "SCC": r'C:\Users\andre\dev\BSc-Project\Dataset\data\val\SCC',
    "SEK": r'C:\Users\andre\dev\BSc-Project\Dataset\data\val\SEK'    
}

class_directories_test = {
    "ACK": r'C:\Users\andre\dev\BSc-Project\Dataset\data\test\ACK',
    "BCC": r'C:\Users\andre\dev\BSc-Project\Dataset\data\test\BCC',
    "MEL": r'C:\Users\andre\dev\BSc-Project\Dataset\data\test\MEL',
    "NEV": r'C:\Users\andre\dev\BSc-Project\Dataset\data\test\NEV',
    "SCC": r'C:\Users\andre\dev\BSc-Project\Dataset\data\test\SCC',
    "SEK": r'C:\Users\andre\dev\BSc-Project\Dataset\data\test\SEK'    
}


# Function to move images to their respective class directories
def move_images(csv_reader, image_dir, class_directories):
    for row in csv_reader:
        filename = row[-2]  # Assuming the filename is in the second to last column
        image_path = os.path.join(image_dir, filename)
        if os.path.exists(image_path):
            image_class = row[-9]  # Assuming the class is int the 9th last column
            if image_class in class_directories:
                destination_dir = class_directories[image_class]
                shutil.move(image_path, destination_dir)
                print(f"Moved {filename} to {image_class} directory")
            else:
                print(f"No directory found for class: {image_class}")
        else:
            print(f"Image not found: {filename}")
```

```python
# Read the CSV file and call move_images function
with open(csv_file, newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    next(csv_reader)  # Skip header if exists
    move_images(csv_reader, train_dir, class_directories_train)

with open(csv_file, newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    next(csv_reader)  # Skip header if exists
    move_images(csv_reader, val_dir, class_directories_val)

with open(csv_file, newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    next(csv_reader)  # Skip header if exists
    move_images(csv_reader, test_dir, class_directories_test)
```
