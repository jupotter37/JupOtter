```python
import numpy as np
import pandas as pd
import sklearn
```

```python
train_data=pd.read_csv("train.csv")
test_data=pd.read_csv("test.csv")
```

```python
train_data.head(10)
```

```python
data = train_data
data = data.append(test_data)
print (len(train_data))
print (len(test_data))
print (len(data))

train_data.info()
print("----------------------------")
test_data.info()
```

```python
ColumnsToDrop = ["PassengerId","Name","Cabin","Ticket"]
data_1 = data.drop(ColumnsToDrop, axis=1)
data_1.head(10)
```

```python
data_1["Sex"] = data_1["Sex"].apply(lambda x: 1 if x == "male" else 0)
data_1.head()
```

```python
data_2 = data_1
data_2["Age"] = data_2["Age"].fillna(data_2["Age"].mean())
data_2.head(10)
```

```python
data_3=data_2
nf = pd.get_dummies(data_3["Embarked"], prefix="em")
for col in nf:
    data_3[col] = nf[col]
data_3 = data_3.drop("Embarked", axis=1)
data_3.head(20)
```

```python
data_5=data_3
nf = pd.get_dummies(data_5["Pclass"], prefix="pclass")
for col in nf:
    data_5[col] = nf[col]
data_5 = data_5.drop("Pclass", axis=1)
data_5.head(100)
```

```python
trainset, testset = data_5[:len(train_data)], data_5[len(train_data):]
print ("Trainset length:", len(trainset))
print ("Testset length:", len(testset))
```

```python
X, y = trainset.drop (["Survived"], axis = 1), trainset["Survived"]
X.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
Xnp = np.array(X)
ynp = np.array(y)
X_train, X_test, y_train, y_test = train_test_split (Xnp, ynp, test_size=0.5)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit (X_train, y_train)
```

```python
print ("Train accuracy:", (model.predict(X_train) == y_train).mean())
print ("Val accuracy:", (model.predict(X_test) == y_test).mean())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier


clf2 = LogisticRegression()
scores2 = cross_val_score (clf2, X, y, cv=10)


clf3 = DecisionTreeClassifier(max_depth=None, min_samples_split=5,
    random_state=0)
scores3 = cross_val_score(clf3, X, y, cv=5)
                               

clf4 = RandomForestClassifier(n_estimators=100, max_depth=None,
    min_samples_split=5, random_state=0)
scores4 = cross_val_score(clf4, X, y, cv=5)
                               

clf5 = ExtraTreesClassifier(n_estimators=100, max_depth=None,
    min_samples_split=5, random_state=0)
scores5 = cross_val_score(clf5, X, y, cv=5)


print ("score", round(sum(scores2)/len(scores2), 3))
print ("score", round(sum(scores3)/len(scores3), 3))
print ("score", round(sum(scores4)/len(scores4), 3))
print ("score", round(sum(scores5)/len(scores5), 3))


```

```python
class myEnsemble:
    def __init__(self, models):
        self.models = models
    def fit(self, X, y):
        for model in self.models:
            model.fit(X, y)
    def predict (self, X):
        preds = [model. predict(X) for model in self.models]
        res = []
        for i in range(len(X)):
            line = [preds[j][i] for j in range (len(preds))]
            res.append(round(sum(line) / len(line)))
        return res
model = myEnsemble([clf2, clf3, clf4, clf5, clf6])
model.fit(X_train, y_train)
r = model.predict(X_test)
print(sum(r == y_test) / len(r == y_test))
```
