```python
def count_summer(row):
    return row.str.count('Summer').sum()/ row.notna().sum()

train['Summer_count'] = train[Seasons].apply(count_summer, axis=1)

def count_winter(row):
    return row.str.count('Winter').sum()/ row.notna().sum()

train['Winter_count'] = train[Seasons].apply(count_winter, axis=1)

def count_fall(row):
    return row.str.count('Fall').sum()/ row.notna().sum()

train['Fall_count'] = train[Seasons].apply(count_fall, axis=1)

def count_spring(row):
    return row.str.count('Spring').sum()/ row.notna().sum()

train['Spring_count'] = train[Seasons].apply(count_spring, axis=1)

train[['Winter_count','Spring_count','Summer_count','Fall_count']].sample(20)
```

```python
train['Has_1_Season'] = (train[['Winter_count','Spring_count','Summer_count','Fall_count']] == 1).any(axis=1)

print(train['Has_1_Season'].value_counts())

train['Missing_Season'] = (train[['Winter_count','Spring_count','Summer_count','Fall_count']] == 0).sum(axis=1)

train['Missing_Season'].value_counts()
```
