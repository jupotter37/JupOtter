```python
def plot_renko(data, brick_size):
    fig = plt.figure(figsize=(20,10))
    fig.clf()
    axes = fig.gca()   
    prev_num = 0
    for index, number in enumerate(data):
#         print(index, number)
        if number == 1:
            facecolor='red'
        elif number == 0:
            facecolor='blue'
        else:
            facecolor='green'

        prev_num += number

        renko = Rectangle(
            (index, prev_num * brick_size+100), 1, brick_size,
            facecolor=facecolor, alpha=0.5
        )
        axes.add_patch(renko)
    plt.xlim(xlims)
    plt.ylim(ylims)
#     plt.show()
```
