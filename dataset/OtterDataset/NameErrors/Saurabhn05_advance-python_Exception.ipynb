#### Error handling for different exceptions
1. Zero Division error
2. Value Errors

```python
def  division3():
    try:
        num = float(input("Please enter numerator : "))
        den = float(input("Please enter denominator : "))
        div = num/den
        print(f"Numerator : {num}, Denominator : {den} , Division : {div:.4f}")
        return div
    except ZeroDivisionError:
        print("Denominator is division cannot be zero")
    except ValueError:
        print("Numerator and Denominator should both be numbers, blank input is not allowed")
```

```python
# num=10,den=3
d1= division3()
print("Done")
```

```python
d1
```

```python
# num=10,den=0
d2 = division3()
print("Done")
```

```python
# num=a
d3=division3()
```

#### Blocks in exception handling
1. try: The code which you try to execute
2. except: if any error or exception occurs then this block will get executed
3. else: if the try block is executed only then else block will get executed
4. finally: Will always be executed for any block is executed

```python
from math import sqrt
```

```python
sqrt(9)
```

```python
sqrt(225)
```

```python
sqrt(-1)
```

```python
def square_root():
    n = float(input("Please enter a positive number : "))
    s = sqrt(n)
    print(f"Square root of given number {n} is {s:.4f}")
    return s
```

```python
# n = 10
s1 = square_root()
print("done")
```

```python
# n = -12
s2 = square_root()
print("done")
```

```python
# n="e"
s3 = square_root()
print("done")
```

```python
def square_root2():
    try:
        n = float(input("Please enter a positive number : "))
        s = sqrt(n)
    except Exception as e:
        print(f"Exception Occured : {e}")
    else:
        print(f"Square root of {n} is {s:.4f}")
    finally:
        print("Finally Block is always executed")
```

```python
# n =25
square_root2()
```

```python
# n = -3
square_root2()
```

```python
# n ="s"
square_root2()
```

### Raising custom errors

```python
def age_validation():
    age  = int(input("Please enter age in years :"))
    if age <= 0 or age >=100:
        raise ValueError("Age cannot be less than 0 or grater than 100")
    else:
        print(f"Valid age entered : {age}")
        return age
```

```python
# age = 25
a1= age_validation()
```

```python
# age = -3
a2= age_validation()
```

```python
# age = 230
a3= age_validation()
```

### File Handling
1. w - write , Create a new file and write the content
2. r - read , Read the file and import the data in python
3. a - append , add something in an existing file

```python
# Save the content in new file using write mode
with open("example.txt","w") as f1:
    f1.write("This is line 1\n")
    f1.write("This is line 2\n")
```

```python
# Read mode to read the content from txt file
with open("example.txt","r") as f2:
    content = f2.read()
```

```python
print(content)
```

```python
# Append mode to add new content in file
with open("example.txt","a") as f3:
    f3.write("This is line 3\n")
    f3.write("Thank you\n")
```
