# Let's run Welch now

```python
w = welch(voxel, fs = TR, nperseg = nperseg, noverlap = noverlap)
```

# Here is our log log Power Spectrum

```python
plt.loglog(w[0],w[1])
```

# let's perform a linear regression on the log values

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.log10(w[0])[1:].reshape((-1, 1))
y = np.log10(w[1])[1:]
print(x)
print(y)
```

```python
model = LinearRegression().fit(x, y)
```

# This is our slope, our $- \beta$

```python
negbeta = model.coef_
beta = negbeta*-1
print(beta)
```

# $ H = (\beta - 1)/2$

```python
H = (beta + 1)/2
print('H = ', H)
```

# Let's plot our regression onto our PSD

```python
y_new = model.predict(x)
```

```python
# plot the results
plt.figure(figsize=(4, 3))
ax = plt.axes()
ax.scatter(x, y)
ax.plot(x, y_new)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')


plt.show()
```

```python
#looking at a whole slice:

example_slice = 'filtered_func_data_clean_run1.nii.gz'
slice_img = nib.load(example_slice, mmap=False)
slice_array = slice_img.get_fdata()
[N1, N2, slice_num, N3] = slice_array.shape
row = np.arange(0,N1)
column = np.arange(0,N2)
#change 3rd element in array to change what slice you want to view
slice_sq = slice_array[:,:,30,:]
imgplot = plt.imshow(slice_sq[:,:,1],cmap='Greys',  interpolation='nearest')
plt.show()
TR = slice_img.header.get_zooms()
TR = TR[3]
print(TR)

def welch_voxel(i,j):
    global TR 
    voxel = (slice_sq[i,j])
    if np.mean(voxel) == 0:
        return None
    else:
        nperseg = math.floor(len(voxel)/8)
        print(nperseg)
        noverlap = math.floor(nperseg/2)
        print(noverlap)
        w = welch(voxel, fs = TR, nperseg = nperseg, noverlap = noverlap)
        x = np.log10(w[0])[1:].reshape((-1, 1))
        y = np.log10(w[1])[1:]
        model = LinearRegression().fit(x, y)
        negbeta = model.coef_
        beta = negbeta*-1
        print(beta)
        H = (beta + 1)/2
        return H

num_cores = multiprocessing.cpu_count()
output = Parallel(n_jobs=num_cores)(delayed(welch_voxel)(i,j) for i in row for j in column)
output = np.array(output) #convert output into numpy array
output = output.astype(np.float64) #convert type object list elements to type float
Hurst_matrix = output.reshape(N1,N2)
Hurst_image = sns.heatmap(Hurst_matrix)

```
