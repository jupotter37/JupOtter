```python
def value_map(col: Sequence, val_map: Mapping) -> Sequence:
    """when a column has mapped values, use this"""
    mapped_seq = [val_map.get(x, x) for x in col]
    new_seq = [np.nan if x == "Replace with blank" else x for x in mapped_seq]
    return new_seq
```
