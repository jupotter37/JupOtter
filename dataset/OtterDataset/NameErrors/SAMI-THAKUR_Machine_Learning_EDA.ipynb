```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
url="https://drive.google.com/file/d/16JuZpn-BOb1UpHHa55oFwI_Jp2LDS2ak/view?usp=drive_link"
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url)
```

```python
sns.countplot(data=df , x='test_result')
```

<h3>older people tends to fail test more than younger people</h3>

<h3>older people have less physical score than young peopl</h3>

```python
fig,ax = plt.subplots(nrows=1  , ncols=2 , figsize=(15,5))
sns.boxplot(x='test_result' , y='age' , data=df , ax=ax[0])
sns.boxplot(x='test_result' , y='physical_score' , data=df , ax=ax[1])
```

```python
fig,ax = plt.subplots(nrows=1  , ncols=2 , figsize=(15,5))
sns.lineplot(x='age' , y='physical_score' , data=df , ax=ax[0])
sns.heatmap(df.corr(),annot=True , ax=ax[1])
sns.pairplot(df,hue='test_result')
```

```python
fig,ax = plt.subplots(nrows=1  , ncols=2 , figsize=(15,5))
sns.scatterplot(x='physical_score',y='test_result',data=df, ax=ax[0])
sns.scatterplot(x='age',y='test_result',data=df, ax=ax[1])
```
