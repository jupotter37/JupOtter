Split numerical from categorical features

```python
ds_copy = ds_raw.copy()

ds_num = ds_copy.drop(columns=['ocean_proximity', 'median_house_value'])
ds_num_np = ds_num.values
n_ex, n_num_feat = ds_num_np.shape

# ds_cat = pd.DataFrame(ds_copy['ocean_proximity'])
ds_cat = ds_copy['ocean_proximity']
ds_lab = ds_copy['median_house_value']
```

Attribute adder for numerical feature

```python
from sklearn.base import TransformerMixin
from sklearn.preprocessing import MinMaxScaler


# Index dictionary to go from NumPy to Dataframe
ix = {feat: ds_num.columns.get_loc(feat) for feat in ds_num.columns}


class AttributeAdder(TransformerMixin):
    """
    Add attributes incercept and bedrooms_per_room.
    """
    
    def __init__(self):
        super().__init__()

    def _normalize(self, feature):
        """
        Normalize feature using MinMaxScaler.
        e.g. all values in between zero and one
        Return normalized feature w shape (-1, 1).
        
        feature: ndarray w shape (-1,)
        """
        assert feature.shape == (n_ex,)
        feature_res = feature.reshape((-1, 1))
        assert feature_res.shape == (n_ex, 1)
        mms = MinMaxScaler()
        norm_feature = mms.fit_transform(feature_res)
        return norm_feature
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        """
        Return new design matrix w/o changing X.
        """
        assert type(X) == np.ndarray
        
        """Normalize longitude and latitude"""
        longitude_norm = self._normalize(X[:, ix['longitude']])
        latitude_norm = self._normalize(X[:, ix['latitude']])
        
        """Add intercept feature
        latitude_norm = - longitude_norm + intercept"""
        intercepts = longitude_norm + latitude_norm
        assert intercepts.shape == (n_ex, 1)

        """Add bedrooms_per_room feature"""
        bedrooms_per_room = X[:, ix['total_bedrooms']] / X[:, ix['total_rooms']]
        bedrooms_per_room = bedrooms_per_room.reshape((n_ex, 1))
        
        """Add room_per_household feature"""
#         rooms_per_household = X[]
        
        new_X = np.hstack([X, intercepts, bedrooms_per_room])

        return new_X
```

Transform data and return design matrix `X`

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, Imputer, LabelBinarizer


numerical_pipeline = Pipeline([
    ('imputer', Imputer(strategy='median')),
    ('attribute_adder', AttributeAdder()),
    ('scaler', StandardScaler())
])

ds_num_tr = numerical_pipeline.fit_transform(ds_num)
ds_cat_tr = LabelBinarizer().fit_transform(ds_cat)  # Use CategoricalEncoder
X = np.hstack([ds_num_tr, ds_cat_tr])
```

```python
from sklearn.model_selection import train_test_split


ss = StandardScaler()
ds_lab_val = ds_lab.values
ds_lab_val_reshaped = ds_lab_val.reshape((-1, 1))
y = ss.fit_transform(ds_lab_val_reshaped)
X_trn, X_tst, y_trn, y_tst = train_test_split(X, y, test_size=.1)
```

## Fit ML models

Evaluate model performance by *K*-fold cross validation: train on *K-1* folds and evaluate performance on the remaining fold. Average score reported.

```python
from sklearn.model_selection import cross_val_score
from sklearn.utils import shuffle

def display_score(model, X, y):
    X, y = shuffle(X, y)
    
    if model == for_reg:
        y2 = y.reshape(-1)
    else:
        y2 = y
        
    scores = cross_val_score(
        model, 
        X, 
        y2,
        scoring='neg_mean_squared_error', 
        cv=3
    )
    # std is 68% confidence interval
    print(str(type(model)))
#     print(scores)
    print('Score: %.2f' % scores.mean(), '+/- %.2f' % scores.std())
#     print('\n')

print('ON TRAINING SET')
display_score(lin_reg, X_trn, y_trn)
display_score(tree_reg, X_trn, y_trn)
display_score(for_reg, X_trn, y_trn)
print('\nON TEST SET')
display_score(lin_reg, X_tst, y_tst)
display_score(tree_reg, X_tst, y_tst)
display_score(for_reg, X_tst, y_tst)
```
