```python
from nltk import word_tokenize,pos_tag
```

```python
def nouns_adj(text):                                                             
    is_noun_adj=lambda pos: pos[:2]=='NN' or pos[:2]=='JJ'
    tokens=word_tokenize(text)
    noun_adj=[x for (x,pos) in pos_tag(tokens) if is_noun_adj(pos)]
    return ' '.join(noun_adj)
```

```python
sent='Ayesha is a very good woman. She is a resident of Jharkhand. She cooks yummy pancakes. Her friends also call her aasha.'
```
