```python
import re
import numpy as np
import sys
import os
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import *
```

```python
df2= pd.read_json("English.jsonl", lines=True)
```

```python
df=pd.DataFrame(df2)
df['ABSTRACT']=df['ABSTRACT'].fillna("").apply(list)
```

```python
df
```

```python
def list_to_string(text):
    abc=''.join(text)
   
    return abc
```

```python
df['Preprocessing'] = df['ABSTRACT'].apply(list_to_string)
```

```python
df
```

## lower text

```python
def lower_text(text):
    text=text.lower()
    return text 
```

```python
df['Preprocessing'] = df['Preprocessing'].apply(lower_text)
```

## remove_punctuations

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
        
```

```python
df['Preprocessing'] = df['Preprocessing'].apply(remove_punctuations)
```

## remove double space

```python
def double_space(text):
    for i in range(0, len(text)):
        text= text.replace(' +', ' ')
    return text
```

```python
df["Preprocessing"] = df['Preprocessing'].apply(double_space)
```

## remove special characters

```python
def remove_special(text):
    ls=[]
    for k in text.split("\n"):
        ls.append(re.sub(r"[^a-zA-Z0-9]+", ' ', k))
    text= " ".join(ls)
    return text
```

```python
df["Preprocessing"] = df['Preprocessing'].apply(remove_special)
```

```python
df
```

## Tokenization

```python

df["Preprocessing"] = df['Preprocessing'].apply(nltk.word_tokenize)
```

## Stopwords

```python
def Stopwords(text):
    stop_words = stopwords.words('english')
    word_tokens = [t for t in text if not (t in stop_words)]
    return word_tokens
```

```python
df["Preprocessing"] = df['Preprocessing'].apply(Stopwords)
```

```python
df
```

## stemming 

```python
def stemming(text):
    ps = PorterStemmer()
    stems=[]
    for i in text:
        stems.append(ps.stem(i))
    return stems
```

```python
df["Preprocessing"] = df['Preprocessing'].apply(stemming)
```
