```python
# Things to import for various reasons
from __future__ import unicode_literals
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from nltk.corpus import stopwords
from collections import OrderedDict
from operator import itemgetter 
import spacy
import numpy
import csv
import json

import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from textblob import TextBlob
from nltk.corpus import wordnet
```

```python
# Things to do only once

# load model, english, used for doc2vec
nlp = spacy.load('en_core_web_lg') 

# Run once, it is used to load file via which the model to make Clusters is trained
doc = []
id_keeper = {}
with open('data_gifs3.csv') as csv_file:
    csv_reader = csv.reader(csv_file)
    line_count = 0
    for row in csv_reader:
        doc.append(row[2])
        id_keeper.update({row[2]:row[0]})
print(doc)
```

```python
# Run only once, it is used to train model that will make Clusters
documents = doc

vectorizer = TfidfVectorizer(stop_words=(set(stopwords.words("english"))  - set(["up"])).union(set(["I"]),set(["gif"]),set(["na"])))
X = vectorizer.fit_transform(documents)

true_k = 1000
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=1000, n_init=2)
model.fit(X)

# Helps to know the tags in each cluster
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :5]:
        print(' %s' % terms[ind]),
    print

print("\n")
print("Prediction")

#Define the puncuations marks
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
```

```python
#Give the lemotized input to form clusters

#Taking input
str_input = "party and party"

# remove punctuations and convert to lower case
text = ""
for char in str_input:
    if char not in punctuations:
        text = text + char.lower()
        
# removing stopwords 
stop_words=(set(stopwords.words("english")))
tokenized_word=word_tokenize(text)

#word tokenizing
filtered_sent=[]
for w in tokenized_word:
    if w not in stop_words:
        filtered_sent.append(w)

lem = WordNetLemmatizer()

# Lemmatize
lemm_sent = []
for word in filtered_sent:
    lemm_sent.append(lem.lemmatize(word,"v"))

print(lemm_sent)
# lemm__sent = ""
# for l in lemm_sent:
#     lemm__sent+=" "+str(l)
# print(lemm__sent)
# Predicting the cluster of the input 
user_input = str(lemm_sent)
Y = vectorizer.transform([str_input])
prediction = model.predict(Y)
print(prediction)
```

```python
#Remove the tags in the related clusters
tags = []
for ind in order_centroids[prediction[0], :5]:
    tags.append(terms[ind])
```

```python
#Load the latest dataset into Dataset

doc2 = []
with open('data_gifs_newest3.json') as json_file:
    json_reader = json.load(json_file)
    for row in json_reader:
        doc2.append(row)
#print(doc2)
```

```python
#Find the relevent tuples from the new dataset which match the tags
relevent_tuples = []
for d in doc2:
    if d['Tag'] in tags:
        relevent_tuples.append(d)
    
```

```python
#checking the length of tuples from clusters
for r in relevent_tuples:
    print(r['description'], r['link'],r['id'])
print(len(relevent_tuples))
```
