```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('CSCI_Survey.csv',delimiter=',')
```

```python
csci_df = pd.read_csv('CSCI_Survey.csv')
csci_df.head(30)
```

```python
#standardized the data within csv file rather than through pandas
```

```python
csci_df.columns
```

```python
age_count= csci_df["How old are you?"].value_counts()

age_count.plot(kind="bar")
plt.xlabel('Age')
plt.ylabel('Number of students')
plt.title('Bar graph of students age')
plt.grid()

#this shows the ages of the students within the class
```

```python
sibling_count = csci_df["How many siblings do you have?"].value_counts()
sibling_count.plot(kind="bar")
plt.xlabel('Number of siblings')
plt.ylabel('Number of students')
plt.title('Bargraph of siblings')
plt.grid()

#this shows how many students have either 1, 2, 3, or 4 siblings in their family
```

```python
age_count = csci_df["How many siblings do you have?"].value_counts()
colors = ['pink','blue', 'red', 'green', 'yellow']
age_count.plot(kind="pie", y = "How old are you?", autopct='%1.1f%%', colors=colors)
plt.title('Pie chart of siblings')
plt.xlabel('Number of siblings')

#this shows the percentage of students with 1, 2, 3, or 4 siblings


```

```python
food_count = csci_df["What is your favorite type of food?"].value_counts()
colors = ['pink','blue', 'red', 'green', 'yellow', 'orange', 'white', 'purple']
food_count.plot(kind="pie", y = "What is your favorite type of food?", autopct='%1.1f%%', colors=colors)
plt.ylabel("")
plt.title("Favorite food of students")

#this is a pie chart showing the breakdown of students favorite food and the percentage of it
```

```python
plt.scatter(csci_df['How old are you?'], csci_df['How much experience do you have programming?'])
plt.ylabel("Experience")
plt.xlabel("Age")
plt.title("Age and Experience of Students in cs")
plt.grid()
#this is a scatter plot of the age of students with their experience in cs.
```
