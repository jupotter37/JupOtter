```python
import pandas as pd
import openai
import random
import requests
import json
```

```python
data = pd.read_csv("./data/icl_tagging.csv",index_col=0)
original_data = pd.read_csv("./data/TruthfulQA.csv",index_col=0)
data.head(1)
```

```python
data
```

```python
result = {}
```

```python
for index, row in data.iterrows():
    if (row['question'] in result.keys()):
        continue
    hint = str(row['url']).replace(row['answer'],"")
    prompt = f"\n\nHuman: Complete the sentence missing in () according to the hint in one word: [{row['question']}]. Hint:[{hint}]\n Reply the answer only in one word without full sentence.\n\nAssistant:"
    print(index)
    url = ""
    headers = {
    "accept": "application/json",
    "anthropic-version": "2023-06-01",
    "content-type": "application/json",
    "x-api-key": "" # Make sure ANTHROPIC_API_KEY is defined
}

# Define the payload data
    data_r = {
        "max_tokens_to_sample": 64,
        "model": "claude-2",
        "prompt": prompt
    }

    # Make a POST request
    response = requests.post(url, headers=headers, json=data_r)
    data_json = json.loads(response.text)
    # Extract content from the dictionary
    content = data_json["completion"].replace("\n","")
    print(content)
    result[row['question']] = content
    
```
