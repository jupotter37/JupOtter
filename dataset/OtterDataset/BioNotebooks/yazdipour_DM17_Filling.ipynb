```python
dff=df[(df['country']=='n/a') & ((df['uniSelected']!='n/a')|(df['uniAccepted']!='n/a'))][['country','uniSelected','uniAccepted']].copy()
print('Count',dff.count()[0])
dff.head()
```

```python
import re
def checkFarsi(w):
    regexp = re.compile(r'[a-zA-Z?]')
    return (not regexp.search(w))
```

```python
col=['uniSelected','uniAccepted']
cdf.name=cdf.name.apply(str)
for c in col:
    changed=0
    for i in dff.index:
        temp=dff.ix[i][c]
        if (checkFarsi(temp) | (temp=='n/a')) :continue
        isin=cdf.name.isin([temp])
        isin=isin[isin==True]
        if len(isin)==0:continue
        inx=isin.index[0]
        res=cdf.ix[inx]['country']
        dff.set_value(i, 'country', res)
        changed+=1
    print(c,'Changed items:',changed)
```

```python
dff.country=dff.country.str.lower()
dff.country.value_counts().head(10)
```

```python
for i in dff.country.index:
    df.set_value(i,'country',dff.country[i])
```
