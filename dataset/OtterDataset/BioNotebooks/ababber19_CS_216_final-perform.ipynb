```python
# This shows the mean statistics of each department.
df.groupby("department_name").sum().sort_values("num_student", ascending=False)

# This shows us how many ratings each department has. We can use this to compare the 
# departments with the most amount of reviews.
```

```python
# This creates a new dataframe that looks at the departments across all schools, the total number of ratings of 
# each department, the average rating of each department, and the average difficulty of each department

df_dep_info = (df.groupby("department_name").sum()["num_student"]).to_frame()

df_dep_info = df_dep_info.merge(df.groupby("department_name").mean()[["student_difficult", "star_rating"]], left_index=True, right_index=True)
df_dep_info = df_dep_info.reset_index()
```
