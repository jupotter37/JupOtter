```python
import pandas as pd
from pathlib import Path
```

```python
keystone_filepath = Path('Resources/2019Keystone_School_cleaned.xlsx')
expenditure_filepath = Path('Resources/2018-2019 Per Pupil Expenditures.xlsx')

keystone2019_df = pd.read_excel(keystone_filepath)
expenditure_df = pd.read_excel(expenditure_filepath)
expenditure_df
```

```python
keystone2019_df.head()
```

```python
expenditure_df = expenditure_df[['AUN', 'LEA Name', 'County', 'Building Name','2018-2019 Local Per Pupil Expenditure']]
expenditure_df.head()
```

```python
keystone2019_df['School Name'] = keystone2019_df['School Name'].str.strip().str.lower()
expenditure_df['Building Name'] = expenditure_df['Building Name'].str.strip().str.lower()
```

```python
merged_df = pd.merge(keystone2019_df, expenditure_df, left_on=['School Name', 'County'], right_on=['Building Name', 'County'], how='inner')
merged_df
```

```python
keystone_expenditure_df = merged_df.drop(['AUN_y', 'LEA Name', 'Building Name'], axis=1)
keystone_expenditure_df = keystone_expenditure_df.rename(columns={'AUN_x': 'AUN'})
keystone_expenditure_df['School Name'] = keystone_expenditure_df['School Name'].str.strip().str.upper()
keystone_expenditure_df
```

```python
keystone_expenditure_df['2018-2019 Local Per Pupil Expenditure'].unique()
```

```python
keystone_expenditure_df.to_excel('Resources/keystone_expenditure.xlsx', index=False)
keystone_expenditure_df.to_json('Resources/keystone_expenditure.json', orient='records')
```
