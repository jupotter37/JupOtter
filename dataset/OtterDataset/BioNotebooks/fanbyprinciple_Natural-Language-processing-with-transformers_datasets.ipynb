### using dataframe functionalities

```python
import pandas as pd

emotion.set_format(type="pandas")
df = emotion["train"][:]
df.head()
```

```python
def label_int2str(row):
    return emotion["train"].features["label"].int2str(row)
df["label_name"] = df["label"].apply(label_int2str)
df.head()
```

```python
import matplotlib.pyplot as plt
df["label_name"].value_counts(ascending=True).plot.barh()
plt.title("Frequency of Classes")
plt.show()
```

```python
df["Words Per Tweet"] = df["text"].str.split().apply(len)
df.boxplot("Words Per Tweet", by="label_name", grid=False,
 showfliers=False, color="black")
plt.suptitle("")
```

# From text to tokens

## character level encoding

```python
text = "tokenizing in nlp is tough."
tokenised_text = list(text)
print(tokenised_text)
```

```python
token2idx = {ch: idx for idx, ch in enumerate(sorted(set(tokenised_text)))}
```

```python
input_ids =  [token2idx[token] for token in tokenised_text]
input_ids
```

```python
token2idx
```

last step is to convert the token into a unique numerical identifier. 


```python
import torch
import torch.nn.functional as F

input_ids = torch.tensor(input_ids)
one_hot_encodings = F.one_hot(input_ids, num_classes=len(token2idx))
one_hot_encodings
```

```python
one_hot_encodings.shape
```

## Word level encoding

```python
tokenized_text = text.split()
tokenized_text
```

```python
winput_ids = { token:id for id, token in enumerate(list(set(tokenized_text)))}
```

```python
winput_ids
```

```python
input_no = [winput_ids[t] for t in tokenized_text]
```

```python
input_no
```

```python
input_ids = torch.tensor(input_no)
one_hot_encodings = F.one_hot(input_ids, num_classes=len(token2idx))
one_hot_encodings
```

```python
from transformers import AutoTokenizer
model_ckpt = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
text
```

```python
encoded_text = tokenizer(text)
print(encoded_text)
```

```python
tokens = tokenizer.convert_ids_to_tokens(encoded_text.input_ids)
print(tokens)
```
