```python
### Read Data Frame from Google Sheet and Convert to Cleaned Up Data Frame

scope = ["https://spreadsheets.google.com/feeds",'https://www.googleapis.com/auth/spreadsheets',"https://www.googleapis.com/auth/drive.file","https://www.googleapis.com/auth/drive"]

creds = ServiceAccountCredentials.from_json_keyfile_name("Service_Account_Creds.json", scope)
print(creds)
```

```python
client = gspread.authorize(creds)

wks = 'AP Calc (Coapman)'

n=0

# print("Sleeping for 100 seconds!")
# time.sleep(100)

for course in courses:
    wrkbk_id = workbooks[course]
    book = client.open_by_key(wrkbk_id)
    for sheet_name in all_sheets[course]:
        sheet = book.worksheet(sheet_name)
        matrix = sheet.get_all_values()
        columns = matrix[0]
        print(sheet_name)
        if n == 0:
            df = pd.DataFrame(matrix, columns = columns)
            df = df.drop(index=0).reset_index().drop(columns = 'index')
            n += 1
        else:
            df_temp = pd.DataFrame(matrix, columns = columns)
            df_temp = df_temp.drop(index=0).reset_index().drop(columns = 'index')
            frames = [df, df_temp]
            df = pd.concat(frames)
    print("Sleeping for 100 seconds!")
    # time.sleep(100)


df.head()
```

```python
### Clean Data Frame to Prepare for SQL

# Melt the Data Frame to Clean Data Frame to list 'Daily Observations'
dates = df.columns[6:20]
df2 = pd.melt(df, id_vars = ['Student ID Number','First Name','Last Name','Grade', 'Teacher', 'Period'], value_vars = dates, var_name = 'date', value_name = 'result')

# Make all column names a single string
df2.columns = ['Student_ID_Number', 'First_Name', 'Last_Name','Grade', 'Teacher', 'Period', 'date','result']

# Convert the Date Column into a datetime format
df2['date'] = df2['date'].apply(lambda x: datetime.strptime(x, '%A, %B %d, %Y'))
df2
```

```python
### Write the Data Frame to the SQL db

# Define function to create connection to database
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return conn
 
# Export the clean dataframe to the SQL Database
def update_observations():
    database = r"C:\users\kyled\documents\github\zoomengagement\zoom_database.db"
    
    # create a database connection
    conn = create_connection(database)
 
    # create tables
    if conn is not None:
        df2.to_sql('observations', con=conn, if_exists = 'replace')
    else:
        print("Error! cannot create the database connection.")

update_observations()
```
