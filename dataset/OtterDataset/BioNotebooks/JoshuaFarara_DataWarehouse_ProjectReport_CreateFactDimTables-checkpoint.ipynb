## Transform Integrated Data to populate data warehouse

```python
pop_Integrated_Data = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\pop_Integrated_Data.csv")
pop_Integrated_Data
```

```python
dim_major = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_major.csv")
dim_degree = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_degree.csv")
dim_gparank = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_gparank.csv")
dim_semester = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_semester.csv")
dim_status = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_status.csv")
dim_address = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_address.csv")
dim_year = pd.read_csv(r"C:\Users\User\Documents\School\FallGrad2023\DW\DataWarehouse_ProjectReport\data\DimensionTables\dim_year.csv")

```

```python
# pop_Integrated_Data['MajorID'] = pd.to_numeric(pop_Integrated_Data['MajorID'], errors='coerce')
transformed_data = pop_Integrated_Data.merge(dim_major, left_on='MajorID', right_on='major_name', how='left')
transformed_data = transformed_data.merge(dim_degree, left_on='DegreeID', right_on='degree_name', how='left')
transformed_data = transformed_data.merge(dim_gparank, left_on='GPARankID', right_on='gparank', how='left')
transformed_data = transformed_data.merge(dim_semester, left_on='SemesterID', right_on='semester_abbreviation', how='left')
transformed_data = transformed_data.merge(dim_status, left_on='StatusID', right_on='status_name', how='left')
transformed_data = transformed_data.merge(dim_address, left_on='AddressID', right_on='address_city', how='left')
transformed_data = transformed_data.merge(dim_year, left_on='YearID', right_on='year', how='left')
transformed_data['semester_id'].fillna(3, inplace=True)
transformed_data['status_id'].fillna(3, inplace=True)
transformed_data['semester_id'] = transformed_data['semester_id'].astype(int)
transformed_data['status_id'] = transformed_data['status_id'].astype(int)
```
