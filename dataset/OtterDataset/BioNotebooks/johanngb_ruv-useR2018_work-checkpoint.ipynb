```r
library(ggplot2)
```

```r
set.seed(12345)
X = matrix(rep(1:2, each=6))
beta = runif(15); beta[11:15] = 0; beta = t(beta)
W = matrix(runif(12) + (1:12)/12)
alpha = runif(15); alpha = t(alpha)
biology  = X%*%beta
unwanted = W%*%alpha
```

```r
matrixplot = function(a, colors, alph=.05)
{
    mat = a[[1]]
    y = rep(nrow(mat):1, ncol(mat))
    x = rep(1:ncol(mat), each=nrow(mat))
    df = data.frame(x=x, y=y, z=as.vector(mat))
    gg = ggplot(df, aes(x=x, y=y)) +
      theme_classic() + 
      theme(axis.line=element_blank(), axis.title=element_blank(), axis.text=element_blank(), axis.ticks=element_blank()) +
      guides(fill=FALSE) +
      geom_raster() +
      aes(fill=z) +
      scale_fill_gradient(low=alpha("black", alph), high=alpha(colors[1], alph))
    if (length(a)>1) for (i in 2:length(a))
    {
        gg = gg +       
          geom_raster() +
          aes(fill=z) +
          scale_fill_gradient(low=alpha("black", alph), high=alpha(colors[i], alph))
    }
    return(gg)
}

matrixplot(list(biology, biology), c("yellow", "blue"))
```

```r
matrixplot = function(a, colors)
{
    mat = a[[1]]
    y = rep(rep(nrow(mat):1, ncol(mat)), length(a))
    x = rep(rep(1:ncol(mat), each=nrow(mat)), length(a))
    z = as.vector(a[[1]])
    group = rep(1:length(a), each=length(z))
    if (length(a)>1) for (i in 2:length(a)) z = c(z, as.vector(a[[i]]))
    df = data.frame(x=x, y=y, z=z, group=as.factor(group))
    df$z = df$z/max(df$z)
    gg = ggplot(df, aes(x=x, y=y)) +
      theme_classic() + 
      theme(axis.line=element_blank(), axis.title=element_blank(), 
            axis.text=element_blank(), axis.ticks=element_blank(),
            panel.grid.major = element_blank(),
            panel.grid.minor = element_blank(),
            panel.border = element_blank(),
            panel.background = element_blank()) +
      guides(alpha=FALSE, fill=FALSE) +
      geom_raster() +
      aes(fill=group, alpha=z) +
      scale_fill_manual(values=colors) +
      geom_tile(colour="white") +
      theme(panel.background = element_rect(fill = 'black')) + 
      coord_cartesian(xlim=c(.5,ncol(mat)+.5), ylim=c(.5,nrow(mat)+.5), expand=FALSE) + 
      scale_alpha_identity()
    return(gg)
}

matrixplot(list(biology, unwanted), c("red", "blue"))
matrixplot(list(biology, unwanted), c("white", "white"))
```

```r
pdf("test.pdf",height=12, width=15)
print(matrixplot(list(biology, unwanted), c("red", "blue")))
dev.off()

png("test.png",height=1200, width=1500)
print(matrixplot(list(biology, unwanted), c("red", "blue")))
dev.off()

svg("test.svg",height=12, width=15)
print(matrixplot(list(biology, unwanted), c("red", "blue")))
dev.off()
```
