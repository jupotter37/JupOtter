## Set Up

```python
# use py file to set up workspace
from src.ros_ale_functions import *

# load data
ros_meta = pd.read_csv('input/ros_ale_p1k_metadata.csv',
                       index_col = 0)
mutations = pd.read_excel('input/S1_ros_ale_mutations.xlsx',
                          index_col = 0)

label_path = 'input/mutation_table_labels/S1_metabolism.csv'
met_labels = pd.read_csv(label_path, index_col = 0)
```

## Generate Figure

Final touches were added in Adobe Illustrator.

```python
fig, ax = plt.subplots(figsize = (5,5), dpi = 140)

met_genes = ['ilvH', 'aceE', 'lacZ', 'cyoB', 'ubiF', 'glnX', 'gltA',
             'sucA', 'aroG', 'mgsA', 'icd', 'ppsA', 'zwf', 'torZ',
             'nuoM', 'nuoG', 'nuoC', 'pta', 'trxC', 'cpdA', 'arcB',
             'arcA', 'coaA']

# prefer to only show missense SNPs in icd
skip_muts = ['e14 deletion', 'icd H366H', 'icd T370T']

ax, mut_table = mut_color_table(mutations, ros_meta, 
                                met_genes, met_labels,
                                ax, skip_muts = skip_muts)
plt.savefig('output/fig_s1_raw.pdf')
```
