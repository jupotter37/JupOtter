```python
subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
marks = [97, 68, 59, 81, 77, 92]
plt.bar(subjects,marks,color='blue')
plt.title("A Bar Graph Showing Relationship Between X & Y")
plt.xlabel("Categorical Data - Subjects")
plt.ylabel("Discrete Values - Marks")
plt.show()
```

```python
subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
student1 = [97, 68, 59, 81, 77, 92]
student2 = [88, 61, 80, 40, 62, 52]
student3 = [54, 62, 77, 54, 71, 98]

index = np.arange(6)
print(index)
width = 0.05

plt.bar(index,student1,width,color='aqua',label='Student 1')
plt.bar(index,student2,width,color='r',label='Student 2')
plt.bar(index,student3,width,color='blue',label='Student 3')

plt.title("Stacked Bar Graph Example")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.xticks(ticks=index + width/2,labels=subjects)

plt.legend()

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
# above x,y are coordinate vectors
X, Y = np.meshgrid(x,y)
# above X,Y are coordinate matrices from coordinate vectors
## Here, linspace returns evenly spaced numbers over a specified interval. 
## meshgrid returns coordinate matrices from coordinate vectors.

## We are generating the sin values of square-root of X**2+Y**2. 
## In the next code snippet, we are using this result 
## to plot along with X and Y values in a 3D plot.

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```python
figure = plt.figure(1,figsize=(12,4))
```

#### The rstride and cstride kwargs set the stride used to sample the input data to generate the graph. If 1k by 1k arrays are passed in, the default values for the strides will result in a 100x100 grid being plotted. Defaults to 10. Raises a ValueError if both stride and count kwargs are provided. If rstride and cstride are not 1, not all the points are used to draw the surface. This saves time, because the more the points to plot, the longer the time needed to compute the plot. But at the same time, if your surface has a high variability, skipping 9 points over 10 in the plot will result in a different picture. [SOURCE :- StackOverFlow]

```python
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X,Y,Z,rstride=1,cstride=1,
                                 cmap=matplotlib.cm.coolwarm,
                                 linewidth=0.1)
plt.show()
```

```python
import math
x = np.arange(0,math.pi*2,0.09)
fig= plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.9,0.9])
ax2 = fig.add_axes([0.62,0.62,0.3,0.3])
y = np.sin(x)
ax1.plot(x,y,'b')
ax2.plot(x,np.cos(x),'r')
ax1.set_title('sine')
ax2.set_title('cosine')
plt.show()
```
