```python
def link_sites(SecondHalf,limita,nume,supernume,urls,csvpath):
    #<-------Colectarea linkurilor de articole------->
    t0 = time.time()
    FirstHalf = 'https://link.springer.com/search/page/'
    count = 0
    viewArticle ='View Article'
    limit = limita
    if(SecondHalf is not None):
        k = SecondHalf
        parent_dir ='C:/Users/40757/Desktop/Desktop/licenta'
        FolderPath = os.path.join(parent_dir, supernume)
        #print(FolderPath)
        if(os.path.isdir(FolderPath)!=True):
            os.mkdir(FolderPath)
        else:
            print('Directory already exists')
        CSVfile = nume + '.csv'
        CSVpath = os.path.join(FolderPath,CSVfile)
        csvpath.append(CSVpath)
        peSite = 0#impunem o limita de pe cate linkuri/website-uri sa colectam date
        result = requests.get(FirstHalf +'1'+k)
        c = result.content
        soup = BeautifulSoup(c,"lxml")
        maxpg = int(soup.find('span',{'class':'number-of-pages'}).get_text().replace(',',''))
        for j in range(1,maxpg):  #j se refera la pagina de articole dintr-o categorie
            print('Page ->' + str(j))
            links = FirstHalf + str(j) + k  #aici se creeaza linkul propiu-zis a paginii de cautare unei categorii
            result = requests.get(links)
            c = result.content
            soup = BeautifulSoup(c,"lxml")
            summary = soup.find('main',{'id':'kb-nav--main'}).find('ol',{'id':'results-list'})#restrangem pe cat posibil locul de cautare
            for row,i in zip(summary.findAll('li',{'class': False}),range(len(summary.findAll('li')))):#class:False elimina linkurile care pointeaza catre jurnale
                if(row.find('p',{'class':'content-type'}).get_text(strip = True) == 'Article'):#selectam doar articolele din toate rezultatele cautarii
                    results = row.find('div',{'class':'actions'}).find_all(string=re.compile('.*{0}.*'.format(viewArticle)), recursive=True)
                    if(len(results)!=0): #verificam daca articolul are continut web sau doar pdf si luam doar cel cu continut web
                        peSite = peSite +1
                        s = row.find('h2').find('a', href = True)# Colectam a doua jumatate a linkului
                        s = 'https://link.springer.com/' + s['href'] #Creeam linkul catre articol
                        count = count+1 #Contorizam
                        print(str(count)+'. '+(s))
                        urls.append(s)
                    if(peSite == limit):
                        break
            if(peSite == limit):
                break
        print("\nThe scrapper collected "+str(peSite)+" links\n")
    else:
        print('This category does not exist ')
    print("TIME TO FINISH FIRST STEP: %.2f seconds " %(time.time()-t0))
    print("\nThere are "+str(maxpg)+" pages full of articles from this field\n")
```

```python
def colectare(urls,categorie,supercategorie,csvpath):
    #Vectorii in care o sa adaugam date
    Vtitle = []
    Vauthors = []
    Vmagazine = []
    Vvolume = []
    Vpages = []
    Vpublicationdate = []
    Vaccesses = []
    Valitmetric = []
    Vlink = []
    Vkeywords = []
    Vcategory = []
    Vsupercategory= []
    Vcontinut=[]

    #<-----Colectarea de date de pe fiecare link de articol------>
    t0 = time.time()
    count = 0
    for url in urls:
        result = requests.get(url)
        c = result.content
        soup = BeautifulSoup(c,"lxml")

        count=count+1

        summary1 = soup.find('div',{'class':'c-article-header'})
        summary2 = soup.find('div',{'class':'c-article-body'})
        if summary1 is None:
            print(str(count)+'Nu o mers')
            continue
            
        Vcategory.append(categorie)
        Vsupercategory.append(supercategorie)
            
        title = summary1.find('h1',{'class':'c-article-title'}).get_text().replace("\\","")
        print(str(count)+'.'+title +'\n')
        Vtitle.append(title)


        sx = summary2.find('div',{'class':'c-bibliographic-information__column'}).find_next_sibling()
        if sx:
            link = sx.find('ul',{'class':'c-bibliographic-information__list'}).find('li',{'class':'c-bibliographic-information__list-item--doi'}).find('span',{'class':'c-bibliographic-information__value'}).get_text()
        else:
            sx = summary2.find('div',{'class':'c-bibliographic-information__column'})
            link = sx.find('ul',{'class':'c-bibliographic-information__list'}).find('li',{'class':'c-bibliographic-information__list-item--doi'}).find('span',{'class':'c-bibliographic-information__value'}).get_text()
        Vlink.append(link)

        authors1 = ''
        for row,i in zip(summary1.find('ul',{'class':'c-article-author-list'}).findAll('li'),range(len(summary1.find('ul',{'class':'c-article-author-list'}).findAll('li')))):
            s = row.find(text=True).get_text() 
            authors1 = authors1 + ', ' + s
        authors = authors1[2:]
        Vauthors.append(authors)

        #necesar if deoarece nu toate articolele au revista
        if(summary1.find('p',{'class':'c-article-info-details'}).find('a',{'data-test':'journal-link'}) is not None):
            magazine = summary1.find('p',{'class':'c-article-info-details'}).find('a',{'data-test':'journal-link'}).get_text()
            Vmagazine.append(magazine)
        else:
            Vmagazine.append('NONE')

        #necesar if deoarece nu toate articolele sunt impartite in volume
        if(summary1.find('p',{'class':'c-article-info-details'}).find('b') is not None):
            volume1 = summary1.find('p',{'class':'c-article-info-details'}).find('b').get_text()
            volume = volume1[7:]
            Vvolume.append(volume)
        else:
            Vvolume.append('NONE')

        #necesar if deoarece la unele articole fac referinta la o pagina, mai multe pagini sau nici una
        if(summary1.find('p',{'class':'c-article-info-details'}).find('span',string='pages ') is not None):
            pages1 = summary1.find('p',{'class':'c-article-info-details'}).find('span',string='pages ').nextSibling
            pages = pages1[:len(pages1)-2]
            Vpages.append(pages)
        elif(summary1.find('p',{'class':'c-article-info-details'}).find('span',string='page ') is not None):
            pages1 = summary1.find('p',{'class':'c-article-info-details'}).find('span',string='page ').nextSibling
            pages = pages1[:len(pages1)-2]
            Vpages.append(pages)
        else:
            Vpages.append('NONE')


        publicationdate = summary1.find('p',{'class':'c-article-info-details'}).find('span',{'data-test':'article-publication-year'}).string
        Vpublicationdate.append(publicationdate)

        #necesar if deoarece nu toate articolele isi contarizeaza accesarile
        searched_word1 = 'Altmetric'
        searched_word2 = 'Accesses'
        results = soup.body.find_all(string=re.compile('.*{0}.*'.format(searched_word2)), recursive=True) #Cautam cuvantul sa vedem daca exista in pagina
        if(len(results)!=0):
            if (summary1.find('div',{'data-test':'article-metrics'}).find('div',{'id':'altmetric-container'}).find('div',{'class':'c-article-metrics-bar__wrapper'}).find('ul',{'class':'c-article-metrics-bar'}).find('li',{'class':'c-article-metrics-bar__item'}).find('p') is not None):        
                accesses1 = summary1.find('div',{'data-test':'article-metrics'}).find('div',{'id':'altmetric-container'}).find('div',{'class':'c-article-metrics-bar__wrapper'}).find('ul',{'class':'c-article-metrics-bar'}).find('li',{'class':'c-article-metrics-bar__item'}).find('p').get_text()
                accesses = accesses1[:len(accesses1)-9] #Pentru a ramane doar cu numarul fara 'Acceses'
                Vaccesses.append(accesses)
            else:
                Vaccesses.append('NONE')
        else:
            Vaccesses.append('NONE')

        #necesar if deoarece nu toate articolele au ALtmetric
        results = soup.body.find_all(string=re.compile('.*{0}.*'.format(searched_word1)), recursive=True) #Cautam cuvantul sa vedem daca exista in pagina
        if(len(results)!=0):
            if(summary1.find('div',{'data-test':'article-metrics'}).find('div',{'id':'altmetric-container'}).find('div',{'class':'c-article-metrics-bar__wrapper'}).find('ul',{'class':'c-article-metrics-bar'}).find('li',{'class':'c-article-metrics-bar__item'}).find_next_sibling() is not None):
                alitmetric1 = summary1.find('div',{'data-test':'article-metrics'}).find('div',{'id':'altmetric-container'}).find('div',{'class':'c-article-metrics-bar__wrapper'}).find('ul',{'class':'c-article-metrics-bar'}).find('li',{'class':'c-article-metrics-bar__item'}).find_next_sibling().find('p').get_text()
                alitmetric = alitmetric1[:len(alitmetric1)-10] #Pentru a ramane doar cu numarul fara  'Alitmetric'
                Valitmetric.append(alitmetric)
            else:
                Valitmetric.append('NONE')
        else:
            Valitmetric.append('NONE')

        #necesar if deoarece nu toate articolele au cuvinte cheie
        keywords = ''
        keywords1 = ''
        if(sx.find('ul',{'class':'c-article-subject-list'}) is not None):
            for row,i in zip(sx.find('ul',{'class':'c-article-subject-list'}).findAll('li'),range(len(sx.find('ul',{'class':'c-article-subject-list'}).findAll('li')))):
                if(len(row.find(text=True).get_text(strip=True))!= 0):
                    s = row.find(text=True).get_text()
                    keywords1 = keywords1 + ', ' + s
                else:
                    s = row.get_text(strip=True)
                    keywords1 = keywords1 + ', ' + s
            keywords = keywords1[2:]
            Vkeywords.append(keywords)
        else:
            Vkeywords.append('NONE')
        references = summary2.find('div',{'id':'MagazineFulltextArticleBodySuffix'})
        author_information = summary2.find('section',{'aria-labelledby':'author-information'})
        ethics = summary2.find('section',{'data-title':'Ethics declarations'})
        additional_info = summary2.find('section',{'data-title':'Additional information'})
        rights_permissions = summary2.find('section',{'data-title':'Rights and permissions'})
        article_info =summary2.find('section',{'aria-labelledby':'article-info'})
        history = summary2.find('section',{'data-title':'Change history'})
        funding = summary2.find('section',{'data-title':'Funding'})
        code = summary2.find('section',{'data-title':'Code Availability'})
        acknowledgements = summary2.find('section',{'data-title','Acknowledgements'})
        if acknowledgements:
            acknowledgements.clear()
        if code:
            code.clear()
        if references:
            references.clear()
        if author_information:
            author_information.clear()
        if ethics:
            ethics.clear()
        if additional_info:
            additional_info.clear()
        if rights_permissions:
            rights_permissions.clear()
        if article_info:
            article_info.clear()
        if history:
            history.clear()
        if funding:
            funding.clear()
        continut = summary2.get_text().replace("\\","")
        Vcontinut.append(continut)

    
    # stocare date 
    dataFrame = pd.DataFrame({
        'SuperCategory':Vsupercategory,
        'Category':Vcategory,
        'Title':Vtitle, 
        'Authors':Vauthors, 
        'Magazine':Vmagazine, 
        'Volume':Vvolume, 
        'Pages':Vpages, 
        'Publication date':Vpublicationdate, 
        'Accesses':Vaccesses, 
        'Alitmetrics':Valitmetric, 
        'Doi link':Vlink,
        'Keywords':Vkeywords,
    })
    print("-----------SUCCESS------------")
    dataFrame.to_csv(csvpath)
    outfile = open("continut.json", "a")
    i=1
    for i in range(len(Vtitle)):
        jsonFrame ={
            'SuperCategory': Vsupercategory[i],
            'Category': Vcategory[i],
            'Title':Vtitle[i],  
            'Keywords':Vkeywords[i],
            'Content': Vcontinut[i]
        }
        json.dump(jsonFrame, outfile)
        outfile.write('\n')
    outfile.close()
    print("TIME TO FINISH SECOND STEP: %.2f seconds" %(time.time()-t0)) #Pentru a vedea performanta aplicatiei
```

```python
#DICTIONAR
count = 0
vect = dict()
for url in urls:
    result = requests.get(url)
    c = result.content
    soup = BeautifulSoup(c,"lxml")
    summary2 = soup.find('div',{'class':'c-article-body'})
    if summary2.findAll('h2',{'class':'c-article-section__title'}):
        for row,i in zip(summary2.findAll('h2',{'class':'c-article-section__title'}),range(len(summary2.findAll('h2',{'class':'c-article-section__title'})))):
            s = row.find(text=True).get_text()
            if(s not in vect):
                vect[s]=1
            else:
                vect[s]=vect[s]+1
    count= count+1
    print(str(count) + '. ')
dictionar = dict(sorted(vect.items(), key=lambda item: item[1],reverse= True))
print(dictionar)
```
