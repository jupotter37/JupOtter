## Data Preparation

```python
dataf=data_frame.replace({'HR':'human resources'}, regex=True)
```

```python
dataf.head(10)
```

```python
dataf['job_title']=dataf['job_title'].str.lower()
dataf.head()
```

```python
data_for_training=dataf.drop(['fit'],axis=1)
data_for_training.head()
```

```python
import string
data_for_training['clean_job_title'] = data_for_training['job_title'].str.replace('\d+', '')
data_for_training.head()
```

```python
#Lemmatize the text

lemmatizer = WordNetLemmatizer()

for i in range(len(data_for_training)):
    tokens =word_tokenize(data_for_training['clean_job_title'][i])
    clean_word= [word for word in tokens if word not in stop_words]
    synonym_word=[]
    for word in clean_word:
        synonym_word.append(word_tokenize(word))
    flat_list = []
    for sublist in synonym_word:
        for item in sublist:
            flat_list.append(item)
    data_for_training['clean_job_title'][i]=TreebankWordDetokenizer().detokenize(flat_list)
```

```python
data_for_training.head()
```

```python
input_phrase = ['Aspiring human resources','seeking human resources']
```

### TFIDF Model

```python

# Create TfidfVectorizer object
vectorizer = TfidfVectorizer()

job_title = data_for_training['job_title']
# Generate matrix of word vectors
tfidf_matrix = vectorizer.fit_transform(job_title)

# Print the shape of tfidf_matrix
print(tfidf_matrix.shape)

# find the tokens from the matrix
features = vectorizer.get_feature_names_out()

print("Length of features:",len(features))

# convering X into an array
tfidf_vector = tfidf_matrix.toarray()
print("The shape of TF-IDF vector is:",tfidf_vector.shape)
```

```python
# lemmatize phrase

def input_vector(phrase):
    lemmatized_phrase = []
    tokenized_phrase = word_tokenize(phrase.lower())
    for word in tokenized_phrase:
        lemmatized_phrase.append(lemmatizer.lemmatize(word))
    lemmatized_phrase = TreebankWordDetokenizer().detokenize(lemmatized_phrase)

    # convert search phrase into a vector
    X1 = vectorizer.transform([lemmatized_phrase])
    X1_vector = X1.toarray()
    return(X1_vector)
```

```python

phrase_vector1=input_vector(input_phrase[0])
phrase_vector2=input_vector(input_phrase[1])
print('Shape of phrase1 vector:',phrase_vector1.shape)
print('Shape of phrase2 vector:',phrase_vector2.shape)
print(len(data_for_training))
```

```python

# compute and print the cosine similarity matrix
cosine_sim = []
for i in range(len(data_for_training)):
    
    tfidf_sim1=cosine_similarity(tfidf_vector[i].reshape(1, -1),phrase_vector1)
    tfidf_sim2=cosine_similarity(tfidf_vector[i].reshape(1, -1),phrase_vector2)
    cosine_sim.append(np.mean(tfidf_sim1.flatten())+np.mean(tfidf_sim2.flatten())/2)

# Add tfidf_similarity in the data frame 
data_for_training['tfidf_similarity'] = cosine_sim

data_for_training['tfidf_similarity'] = data_for_training['tfidf_similarity'].astype(float)

final_output = data_for_training.sort_values(by=['tfidf_similarity'], ascending=False)

print(final_output.head(10))
```
