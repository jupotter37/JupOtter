```python
import os
import requests

def scraping(resource_file, local_dir):
    if not os.path.exists(local_dir):
        os.makedirs(local_dir)

    with open(resource_file) as f:
        urls = f.read().splitlines()

    for i in urls:
        file_name = i.split("/")[-1]
        file_path = os.path.join(local_dir, file_name)
        r = requests.get(i)
        with open(file_path, 'wb') as f:
            f.write(r.content)
        print(file_name)
        


scraping("urls.txt", "/users/iamns45/Downloads/final_articles")
```

```python
import os
from bs4 import BeautifulSoup
import json

data = []
local_dir = "/users/iamns45/Downloads/final_articles"
for filename in os.listdir(local_dir):
    file_path = os.path.join(local_dir, filename)
    if filename == ".ipynb_checkpoints":
        continue
    with open(file_path, 'r') as f:
        soup = BeautifulSoup(f)
        try:
            title = soup.find('title').text
        except AttributeError:
            title = ''
        main_text = soup.find('div', {'id': 'bodyContent'}).text
        for script in soup.find_all("script", type="application/ld+json"):
            ld = json.loads(script.text)
        file_data = {
            'title': title,
            'main_text': main_text,
            'name': ld.get('name', ''),
            'url': ld.get('url', ''),
            'datePublished': ld.get('datePublished', ''),
            'headline': ld.get('headline', ''),
        }
        data.append(file_data)
print(data)
```

```python
import pandas as pd

df=pd.DataFrame(data).set_index("title")

print(df)
```

```python
import re

def preprocess_text(text):
    
    text = re.sub(r'^From Wikipedia, the free encyclopedia.*\n', '', text, flags=re.MULTILINE)
    
    text = re.sub(r'^Retrieved from.*\n|^(Category|Categories):.*\n', '', text, flags=re.MULTILINE)
    
    text = re.sub(r'\[.*?\]', '', text)
    
    text = text.replace('\n', '')
    
    return text


df["main_text"] = df["main_text"].apply(preprocess_text)

df
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

# Define a function to extract various features from text
def extract_features(text):
    doc = nlp(text)
    tokens = [token for token in doc if not token.is_punct]
    nr_tokens = len(tokens)
    nr_characters = len(text)
    nouns = '|'.join([token.text for token in doc if token.pos_ == 'NOUN'])
    adjectives = '|'.join([token.text for token in doc if token.pos_ == 'ADJ'])
    verbs = '|'.join([token.text for token in doc if token.pos_ == 'VERB'])
    lemmas = '|'.join(set([token.lemma_ for token in doc if not token.is_punct]))
    nav = '|'.join(set([token.lower_ for token in doc if not token.is_punct and (token.pos_ == 'NOUN' or token.pos_ == 'ADJ' or token.pos_ == 'VERB')]))
    entities = '|'.join([ent.text for ent in doc.ents])
    noun_chunks = '|'.join([chunk.text for chunk in doc.noun_chunks])
    no_tokens = len(doc)
    no_sentences = len(list(doc.sents))
    no_noun_chunks = len(list(doc.noun_chunks))
    return nr_tokens, nr_characters, nouns, adjectives, verbs, lemmas, nav, entities, noun_chunks, no_tokens, no_sentences, no_noun_chunks

# Add the new columns to the dataframe
df['nr_tokens'], df['nr_characters'], df['nouns'], df['adjectives'], df['verbs'], df['lemmas'], df['nav'], df['entities'], df['noun_chunks'], df['no_tokens'], df['no_sentences'], df['no_noun_chunks'] = zip(*df['main_text'].apply(extract_features))
```
