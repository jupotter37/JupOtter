```python
def factorial(n):
    x = 1
    for i in range(n):
        x*=i+1
    return x
n = int(input("Masukan Bilangan"))
print("Faktorial dari bilang yan di input user adalah",factorial(n))
```

```python
def lingkaran(r):
    luas_lingkaran = 3.14 * r * r
    keliling_lingkaran = 2 * 3.14 * r
    print("Luas lingkaran =", luas_lingkaran)
    print("Keliling Lingkaran =", keliling_lingkaran)
    return luas_lingkaran, keliling_lingkaran

def persegi(s):
    luas_persegi = s * s
    keliling_persegi = 4 * s
    print("Luas Persegi =", luas_persegi)
    print("Keliling Persegi =", keliling_persegi)
    return luas_persegi, keliling_persegi

def segitiga(a, t):
    luas_segitiga = 0.5 * a * t
    print("Luas Segitiga =", luas_segitiga)
    return luas_segitiga

while True:
    print("Jenis bangun datar")
    print("1. lingkaran\n2. persegi\n3. segitiga")
    masukan_rumus = input("Masukkan jenis bangun datar yang ingin dihitung (lingkaran/persegi/segitiga): ")
    
    if masukan_rumus.lower() == "selesai":
        break
        
    if masukan_rumus.lower() in ["lingkaran", "persegi", "segitiga"]:
        try:
            if masukan_rumus.lower() == "lingkaran":
                r_input = input("Masukkan jari-jari lingkaran: ")
                if not r_input.strip():
                    raise ValueError("Nilai tidak boleh kosong")
                if not r_input.strip().isdigit():
                    raise ValueError("Hanya boleh memasukkan angka")
                r = int(r_input)
                print("Jika jari-jari lingkaran =", r, "maka hasilnya adalah", lingkaran(r))
                
            elif masukan_rumus.lower() == "persegi":
                s_input = input("Masukkan sisi persegi: ")
                if not s_input.strip():
                    raise ValueError("Nilai tidak boleh kosong")
                if not s_input.strip().isdigit():
                    raise ValueError("Hanya boleh memasukkan angka")
                s = int(s_input)
                print("Jika sisi persegi adalah =", s, "maka hasilnya adalah", persegi(s))
                
            elif masukan_rumus.lower() == "segitiga":
                a_input = input("Masukkan alas segitiga: ")
                t_input = input('Masukkan tinggi segitiga: ')
                if not a_input.strip() or not t_input.strip():
                    raise ValueError("Nilai tidak boleh kosong")
                if not a_input.strip().isdigit() or not t_input.strip().isdigit():
                    raise ValueError("Hanya boleh memasukkan angka")
                a = int(a_input)
                t = int(t_input)
                print("Jika alas segitiga adalah =", a, "dan tingginya adalah =", t, "maka luas segitiga adalah", segitiga(a, t))
                
        except ValueError as e:
            print(f"Error: {e}. Harap masukkan nilai yang valid")
    else:
        print("Pilihan tidak valid")
```
