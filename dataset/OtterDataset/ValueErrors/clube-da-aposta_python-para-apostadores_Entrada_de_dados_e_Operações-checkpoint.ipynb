## Operações básicas

### Soma - `+`

```python
numero1 = int(input('Digite um número: '))
numero2 = int(input('Digite outro número: '))
soma = numero1 + numero2
print('A soma de', numero1, 'com', numero2, 'é igual a', soma)
```

```python
'Aula ' + '3' + ' do ' + 'Curso de Python'
```

```python
nome = input('Digite o seu nome: ')
idade = input('Digite a sua idade: ')
print('Você se chama ' + nome + ' e tem ' + idade + ' anos.')
```

### Subtração -  `-`

```python
numero1 = int(input('Digite um número: '))
numero2 = int(input('Digite outro número: '))
subtracao = numero1 - numero2
print('A diferença entre', numero1, 'e', numero2, 'é igual a', subtracao)
```

```python
'Aula' - 'A'
```

### Multiplicação - `*`

```python
valor = float(input('Digite um valor a ser apostado: '))
odd = float(input('Digite a odd do evento: '))
resultado = valor * odd
print('\nVocê apostou R$' + str(round(valor,2)) + ' em uma odd de ' + str(round(odd,2)))
print('Ganhando a aposta, você receberá: R$' + str(round(resultado,2)))
print('O seu lucro será de R$' + str(round(resultado - valor,2)))
```

```python
'Clube da Aposta ' * 5 
```

```python
palavra = input('Digite uma palavra: ')
qtd = int(input('Digite a quantidade de vezes que você quer repetir a palavra: '))
print(palavra * qtd)
```

### Divisão - `/`

```python
odd = float(input('Digite a odd do evento: '))
probab = (1 / odd) * 100 
print('A probabilidade do evento ocorrer é de: ' + str(round(probab,2)) + '%')
```

### Exponenciação - `**`

```python
numero = int(input('Digite um número inteiro: '))
expoente = int(input('Digite o número do expoente: '))
resultado = numero ** expoente
print('O número', numero, 'elevado a ' + str(expoente), 'é igual a', resultado)
```

### Divisão inteira - `//`

```python
num1 = int(input('Digite um número qualquer: '))
num2 = int(input('Digite um outro número: '))
resultado = num1 // num2
print('A divisão inteira entre', num1, 'e', num2, 'é igual a', resultado)
```

### Módulo - `%`

```python
num1 = int(input('Digite um número inteiro: '))
num2 = int(input('Digite outro número inteiro: '))
resto = num1 % num2
print('O resto da divisão entre ' + str(num1) + ' e ' + str(num2) + ' é ' + str(resto))
```

**Exercício 1:** Calculadora - Faça um código que solicite ao usuário que digite dois número e exiba na tela o resultado do cálculo utilizando todas as operações vistas durante essa aula.

**Exercício 2:** Thiago fará uma aposta de R$215,00 no jogo Fluminense x Botafogo que irá ocorrer no Campeonato Carioca. Como ele é botafoguense desde criança, decidiu apostar no seu time de coração. Calcule as probabilidades dele ganhar e perder a aposta de acordo com as odds da tabela abaixo, em caso de acerto qual será o montante ganho e o lucro da operação?<br> Exiba na tela os resultados.

| Resultado possível   | Odd  |
|----------------------|------|
| Fluminense           | 2.25 |
| Empate               | 3.30 |
| Botafogo             | 3.20 |

```python
valor = 215.00
odd_fluminense = 2.25
odd_empate = 3.30
odd_botafogo = 3.20

prob_fluminense = round((1 / odd_fluminense) * 100, 2)
prob_empate = round((1 / odd_empate) * 100, 2)
prob_botafogo = round((1 / odd_botafogo) * 100, 2)

montante_botafogo = round((odd_botafogo - 1) * valor, 2)

print('A probabilidade de Thiago ganhar a aposta é de ' + str(prob_botafogo) + '%')
print('A probabilidade de Thiago perder a aposta é de ' + str(prob_fluminense + prob_empate) +'%')
print('Caso a aposta de Thiago seja vencedora ele receberá R$' + str(montante_botafogo))
print('Assim, Thiago teria lucro de R$' + str(montante_botafogo - valor))
```

**Exercício 3:** O *Valor Esperado* é algo com o qual os apostadores e traders devem estar sempre atentos. Ele representa o montante que alguém pode ganhar ou perder se fizesse uma aposta nas mesmas probabilidades repetidamente. [Pinnacle - Como calcular o Valor Esperado](https://bit.ly/2TeKeRw).<br>
Para calcular o Valor Esperado, deve-se utilizar a seguinte equação:

\begin{equation*}
VE   = (Probabilidade\:de\:Ganhar \times Montante\:ganho\:por\:aposta) - (Probabilidade\:de\:Perder \times Montante\:perdido\:por\:aposta)
\end{equation*}
<br>
Agora, calcule o valor esperado da aposta do Thiago segundo os dados apresentados no *Exercício 2*



```python
prob_fluminense = prob_fluminense / 100
prob_empate = prob_empate / 100
prob_botafogo = prob_botafogo / 100

ve = (prob_botafogo * montante_botafogo) - ((prob_empate + prob_fluminense) * valor)
print('O valor esperado para essa aposta é R$' + str(round(ve,2)))
```

Roi é o retorno do ivestimento e ele é calculado utilizando-se a seguinte fórmula:<br>

Agora, calcule o valor total apostado e o lucro ou prejuízo que Thiago obteve nesse específico mês. Após isso, calcule o retorno do investimento que ele obteve.
