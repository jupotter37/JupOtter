```python
import math

def isPrime(n):
    if n <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    if n == 2:
        return True  # 2 is the only even prime number
    if n % 2 == 0:
        return False  # Eliminate even numbers greater than 2

    # Check divisibility from 3 up to the square root of n
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    
    return True
```

```python
def isPalindrome(n):
    def reverse(n):
        rev = 0
        while(n!=0):
            rev = rev * 10 + (n%10)
            n//=10
        return rev
    myRev = reverse(n)
    if(myRev == n):
        return True
    return False
# __main__ segment
while(True):
    n = int(input('Enter n : '))
    print(isPalindrome(n))
```
