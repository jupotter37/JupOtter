#Feature Engineering

```python
final_df['Date_of_Journey'].str.split('/')
```

```python
#we'll have to convert string date of jouney to datetime and generate date, month and year fields
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.head()
```

```python
final_df.drop(columns = ['Date_of_Journey'], inplace = True)
```

```python
final_df.info()
```

```python
#Arrival time

final_df["Arrival_Time"] =  final_df["Arrival_Time"].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df["Arrival_Hour"] =  final_df["Arrival_Time"].str.split(':').str[0]
final_df['Arrival_Minutes']  = final_df["Arrival_Time"].str.split(':').str[1]
```

```python
final_df["Arrival_Hour"] = final_df["Arrival_Hour"].astype(int)
final_df['Arrival_Minutes'] = final_df['Arrival_Minutes'].astype(int)
```

```python
final_df.drop(columns  = ['Arrival_Time'], inplace = True)
```

```python
final_df.head(1)
```

```python
#departure time
final_df["Dep_Hour"] =  final_df["Dep_Time"].str.split(':').str[0]
final_df['Dep_Minutes']  = final_df["Dep_Time"].str.split(':').str[1]
```

```python
final_df["Dep_Hour"] = final_df["Dep_Hour"].astype(int)
final_df['Dep_Minutes'] = final_df['Dep_Minutes'].astype(int)
```

```python
final_df.drop(columns  = ['Dep_Time'], inplace = True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'].isna().sum()
```

```python
final_df[final_df['Total_Stops'].isna()]
```

```python
#either delete this record or lets just assume that maybe 1 stop will be there on the basis of GK
#let us now map the no. of stops in string format to ints

final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop': 0, '2 stops':2, '1 stop':1, '3 stops':3 , '4 stops': 4, 'NaN': 1})
```

```python
#we need only one either route or no. of stops
#so let us now drop route
final_df.drop(columns = ['Route'], inplace = True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
#we can probably just encode additional info

#now let us deal with Duration convert it entirely into minutes

final_df['Duration_Hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]

```

```python
final_df.head()
```

```python
final_df['Duration_Hour'] = final_df['Duration_Hour'].astype(int)
```

```python
final_df[(final_df['Duration']=='5m')] #now this clearly seems impossible so let us just drop these two records
```

```python
final_df.drop([6474, 13343], inplace = True)
```

```python
final_df['Duration_Hour'] = final_df['Duration_Hour'].astype(int)
```

```python
final_df.head(1)
```

```python
final_df['Duration_Minutes'] =final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration_Minutes'].isna().sum()
```

```python
final_df['Duration_Minutes'] = final_df['Duration_Minutes'].fillna('0')
```

```python
final_df['Duration_Minutes'] = final_df['Duration_Minutes'].astype(int)
```

```python
final_df['Duration'] = final_df['Duration_Hour']*60 + final_df['Duration_Minutes']
```

```python
final_df.head()
```

```python
final_df.drop(columns = ['Duration_Hour', 'Duration_Minutes'], inplace = True)
```

```python
final_df.head(1)
```

```python
#categorical features
# we can use labelencoding
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown="ignore")
```

```python
df_Airline = ohe.fit_transform(final_df['Airline'].values.reshape(-1, 1)).toarray()
df_Source = ohe.fit_transform(final_df['Source'].values.reshape(-1, 1)).toarray()
df_Destination = ohe.fit_transform(final_df['Destination'].values.reshape(-1, 1)).toarray()
df_Additional_Info = ohe.fit_transform(final_df['Additional_Info'].values.reshape(-1, 1)).toarray()

```

```python
dfOneHot_Airline = pd.DataFrame(df_Airline, columns=["Airline_" + str(int(i)) for i in range(df_Airline.shape[1])])
dfOneHot_Source = pd.DataFrame(df_Source, columns=["Source_" + str(int(i)) for i in range(df_Source.shape[1])])
dfOneHot_Destination = pd.DataFrame(df_Destination, columns=["Destination_" + str(int(i)) for i in range(df_Destination.shape[1])])
dfOneHot_Additional_Info= pd.DataFrame(df_Additional_Info, columns=["Additional_Info_" + str(int(i)) for i in range(df_Additional_Info.shape[1])])
```

```python
final_df = pd.concat([final_df, dfOneHot_Airline], axis=1)
final_df = pd.concat([final_df, dfOneHot_Source], axis=1)
final_df = pd.concat([final_df, dfOneHot_Destination], axis=1)
final_df = pd.concat([final_df, dfOneHot_Additional_Info], axis=1)
```

```python
final_df.drop(['Airline'], axis=1, inplace = True)
final_df.drop(['Source'], axis=1, inplace = True)
final_df.drop(['Destination'], axis=1, inplace = True)
final_df.drop(['Additional_Info'], axis=1, inplace = True)
```
