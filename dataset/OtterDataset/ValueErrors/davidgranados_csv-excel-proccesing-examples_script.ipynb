```python
import pandas as pd
import requests
from datetime import datetime


pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 2)
CREATE_NEW_DOCUMENT_CITY_ENDPOINT = (
    "https://challengerco.vtexcommercestable.com.br/api/dataentities/CI/documents"
)
API_APP_KEY_NAME = ""
API_APP_KEY_VALUE = ""
API_APP_TOKEN_NAME = ""
API_APP_TOKEN_VALUE = ""
```

```python
"""
Leer csv cities master
"""
cities_master_csv_df = pd.read_csv("source-files/cities-master.csv", encoding="utf-8")
cities_master_csv_df
```

```python
"""
Filtrar solo por las columnas que pertenecen a una especificación
"""
valid_columns = [col for col in cities_master_csv_df.columns if not str(col).startswith("Unnamed")]
cities_master_new_df = cities_master_csv_df[valid_columns]
cities_master_new_df
```

```python
api_request_headers = {
    "accept": "application/json",
    "content-type": "application/json",
    API_APP_KEY_NAME: API_APP_KEY_VALUE,
    API_APP_TOKEN_NAME: API_APP_TOKEN_VALUE,
}
for index, row in cities_master_new_df.iterrows():
    try:
        body = {
            "department": row["department"],
            "city": row["city"],
            "zip": int(row["zip"]),
        }
        print(body)
        api_response = requests.request(
            "POST",
            CREATE_NEW_DOCUMENT_CITY_ENDPOINT,
            headers=api_request_headers,
            json=body
        )
        print(api_response)
    
    except TypeError as e:
        print(e)
```
