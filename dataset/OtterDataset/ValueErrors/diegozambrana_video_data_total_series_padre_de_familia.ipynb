```python
import pandas as pd
import requests
import json
from pyquery import PyQuery as pq
```

```python
URL_TOTAL = "https://totalseries.org/rick-y-morty-espanol-latino"
HEADERS = {'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) "
                         "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50"
                         ".0.2661.102 Safari/537.36"}
data = []
```

```python
main_page_response = requests.get(URL_TOTAL, headers=HEADERS)
main_page_response
```

```python
content = pq(main_page_response.content).find('.entry-content a')
content = content[::-1]
len(content)
```

```python
for index, link in enumerate(content):
    print(index, link.text_content(), link.attrib['href'])
```

```python
link = content[186]
print(link.text_content())
chapter_response = requests.get(link.attrib['href'], headers=HEADERS)
title = pq(chapter_response.content).find('.entry-title')[0].text_content()
description = pq(chapter_response.content).find('div span')[1].text_content()
sources = list(map(lambda x: x.attrib['src'], pq(chapter_response.content).find('.entry-content iframe')))
text = link.text_content()
text = text.replace('x ', '×')
data_pelicula = {
    'title': title,
    # 'season': int(text.split(' ')[0].split('×')[0]) or 'none',
    # 'chapter': 32,# int(text.split(' ')[0].split('×')[1]) or 'none',
    'description': description,
    'sources': sources,
}
print(data_pelicula)
# data.append(d)
# print(d)
# data
```

```python
for link in content:
    print(link.text_content())
    chapter_response = requests.get(link.attrib['href'], headers=HEADERS)
    title = pq(chapter_response.content).find('.entry-title')[0].text_content()
    if ' – ' in title:
        title = title.split(' – ')[1]
    description = pq(chapter_response.content).find('div span')[1].text_content()
    sources = list(map(lambda x: x.attrib['src'], pq(chapter_response.content).find('.entry-content iframe')))
    data.append({
        'title': title,
        'season': int(link.text_content().split(' ')[0].split('×')[0]) or 'none',
        'chapter': int(link.text_content().split(' ')[0].split('×')[1]) or 'none',
        'description': description,
        'sources': sources,
    })
print(data)
```
