```python
#adicionando colunas
df['Ano_de_analise']= 2023

#adicionando coluna com a condição no dataframe

import numpy as np

#colocando

df['Cobrança_abusiva']= np.where(
    df['charges'] > 20000,
    "Sim",
    "não"
)

df.head() #exibindo
```

```python
df.drop(columns='Ano_de_analise') #3xcluir coluna
```

```python
#para excluir realmente precisamos utilizar o inplace

df.drop(columns= 'Ano_de_analise', inplace= True) #excluir permanentemente
df.head()
```

```python
#modificar valores e textos dataframe usando map

#trocaremos sim por 1 e não por zero, sendo assim:

df['Cobrança_abusiva'] = df['Cobrança_abusiva'].map({
    "Sim":1,
    "Não":0
})

#criando coluna para o ano de 2023
df['Ano_Analise']=2023

#atualizando valor de 2023 para 2022

df['Ano_Analise'] = 2022

df.head()
```

```python
#analisando Dados
#é otima para variaveis numericas
df.describe()
```

```python
#para variaveis categoricas podemos usar

df['sex'].value_counts()
```

```python
#para variaveis categoricas podemos usar normalize = True para obter o percentual

df['sex'].value_counts(normalize=True)
```

```python
# A média de idade é diferente de homem para mulher?

df.groupby('sex')['age'].mean()
```

```python
# média da idade entre fumantes e não fumantes
df.groupby('smoker')[['age','charges']].mean()
```

```python
#desvio padrão por sexo e fumante
df.groupby(['smoker', 'sex'])['charges'].std()
```
