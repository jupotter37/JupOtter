- Робота з циклами
> `while` - умовний цикл - ітерації будуть виконуватись, поки умова є істиною (може бути безмежною)

> `for` - із заданою кількістю інтерацій (повторів)

```python
from random import randint
n = ["Богдан", "Іван", "Василь"]
h = ["футбол", "аніме", "програмування"]

j = 0
while j < 3:
    print(">>>{name} любить {hobbi}".format(name=n[j], hobbi=h[randint(0,2)]))
    j+=1

print("="*10, "Це вже не входить в цикл, цикл - закінчився. \nЦикл While не дуже підходить для визначення кількості інтерації ")

# навіть для for можна йти по-різному
for j in range(len(n)):
    print(">>>{name} любить {hobbi}".format(name=n[j], hobbi=h[randint(0,2)]))

print("="*10, "Це завершення виконається поза циклом")

print(f"Демонструємо, що буде виконувати функція range: {list(range(3))}")

# варіант 2 для for (це вже правильне використання for)
for j in n:
    print(">>>{name} любить {hobbi}".format(name=j, hobbi=h[randint(0,2)]))

# спростимо код ще більше
print("="*10)
for j in n:
    print(f">>> {j} любить {h[randint(0,2)]}") 
```

- Знайомство з обробками винятків

```python
# код призведе до помилки бо ми хочемо викликати метод притаманний лише для інт змінних інших класів
ll = [int(2), int(100000), str("10"), float(5000), bytes(1)]

for a in ll:
    print(f"Змінна а має тип {type(a)}.")
    # вкористання блоку IF для обробки помилок є неправильним! 
    if isinstance(a, int): # type(a) == int Це теж саме
        print(f"Кількість байт у змінній a={a.bit_count()}. Вивід через IF.")
    else:
        print("Змінна а не є типу інт!")
    # для обробки помилок ми повинні використовувати конструкцію try
    try:
        print(f"Кількість байт у змінній a={a.bit_count()}. Вивід через TRY.")
    except AttributeError as e:
        print(f"Виникла помилка, деталі: {e}")
        try:
            print(f"Кількість байт у змінній a={int(a).bit_count()}. Вивід через TRY.")
        except ValueError as e:
            print(f"Виникла помилка, деталі: {e}\nТут неможливо порахувати кількість байт у змінній.")
```

> Випадково генеруємо числа де може попадатись число 0, на яке не можна ділити

> За допомогою конструкції try ми можемо опрацювати помилку ділення на 0 та просто задати іншу логіку виконання даної операції

```python
li = [randint(-5,5) for _ in range(30)]
print(li)
for i in li:
    try:
        d = 5/i  
    except ZeroDivisionError:
        print(f"!!! Було згенеровано число {i} на яке ділити не можна!!!")
        d = None
    finally:
        print(f"Будемо ділити 5 на число {i}, результат буде: {d}.")
```

- Знайомство з `Функціями` та `Лямбдами`
> Існують іменні та анонімні/безіменні функції, і лямбда є безіменною!

> Запитались у ChatGPT пояснення, що таке функції:

У мові програмування, `функція (або метод)` - це відрізок коду, який виконує певну операцію або послідовність операцій і може бути викликаний (використовуваний) з інших частин програми. Функції є фундаментальними будівельними блоками в програмуванні і вони допомагають структурувати код, робити його більш читабельним і підтримуваним. 

`Основні характеристики функцій включають`:

1. **Ім'я:** Кожна функція має унікальне ім'я, яке ідентифікує її в програмі.

2. **Параметри (аргументи):** Функція може приймати нуль, один або більше параметрів (аргументів), які використовуються для передачі даних в функцію.

3. **Тіло функції:** В тілі функції зазвичай знаходиться послідовність інструкцій, які визначають операції, які функція виконує.

4. **Значення повернення:** Функція може повертати певне значення (або нічого) як результат її виконання.

5. **Виклик функції:** Щоб використовувати функцію, ви повинні її викликати, передаючи необхідні аргументи, і вона виконає операції і, можливо, поверне результат.

`Основні переваги використання функцій включають`:

- **Модульність:** Функції дозволяють розділити великі програми на менші, легше керовані частини, що сприяє більшому розподілу роботи та підтриму.

- **Повторне використання:** Функції можна використовувати багатократно в різних частинах програми або навіть в інших програмах.

- **Читабельність:** Використання функцій робить код більш зрозумілим, оскільки окремі функції можуть бути документовані та тестовані окремо.

- **Помилки та відлагодження:** Функції полегшують виявлення та виправлення помилок, оскільки ви можете тестувати кожну функцію окремо.

Функції є важливою складовою будь-якої мови програмування і дозволяють програмістам створювати складні програми шляхом розбиття їх на більш прості та керовані частини.

```python
# після def, буде назва функції, за якою ми її викликали
def name_of_my_fun(arg:str):
    # тіло функції, все що буде на рівні відступу, буде виконуватись у функції
    print(f"У функцію було передано значення: {arg}")
    return True if arg != None else False
    # Після цієї команди, більше ніякі операції у функції виконуватись не будуть
    print("Це ніколи не виконається")

print(f"Виклик функції з аргументами: {name_of_my_fun(1)}")

# Функція без аргументів
def no_args_fun():
    return True

print(f"друга функція повернула значення: {no_args_fun()}")
```

```python
# Щодо виклику функції, він здійснюється за допомогою ()
a = no_args_fun

print(a)

def function_call_function(f):
    print("У цю функцію, ми передали інакшу функцію і тепер викликаємо її тут:")
    f()
    print("Передана функція буде викликом.")

def internal_no_args():
    print("Ця функція буде передана кудись.")

function_call_function(internal_no_args)
```

> А тепер спробуємо створити Лямбду

```python
# Протопитом Лямбди буде така функція
def lambda_prototype(a:int, b:float):
    return f"Множимо два отримані числа: {a * b}"

a = lambda_prototype #присвоюємо НЕ викликану функцію у змінну

print(f"Проста функція {a}, далі йде її виклик:", a(2, 2.5))

# Перепишемо цю функцію через Лямбду
lb = lambda a, b: f"Множимо два отримані числа: {a * b}"
print(f"Ми створили Лямбду {lb}, далі йде її виклик:", lb(2, 2.5))
```

```python
# Приклад використання Лямбд
la = lambda i: [i+j if i > 10 else i-j for j in range(i)]

la(13)
```
