```python
path = []
used = [False,False,False,False]

def kfc(n):
    # 기저 조건(종료 조건)
    if len(path) == 4:
        print(*path)
        return
    
    for i in range(1,5):
        if used[i-1] == True: continue
        used[i-1] = True
        path.append(i)
        kfc(n+1)
        path.pop()
        used[i-1] = False


kfc(1)
```
