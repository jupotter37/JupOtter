```python
class Node:
    def __init__(self, id, nama, usia, tinggi_badan, berat_badan):
        self.id = id
        self.nama = nama
        self.usia = usia
        self.tinggi_badan = tinggi_badan
        self.berat_badan = berat_badan
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, id, nama, usia, tinggi_badan, berat_badan):
        new_node = Node(id, nama, usia, tinggi_badan, berat_badan)
        if self.root is None:
            self.root = new_node
        else:
            current = self.root
            while True:
                if id < current.id:
                    if current.left is None:
                        current.left = new_node
                        break
                    else:
                        current = current.left
                else:
                    if current.right is None:
                        current.right = new_node
                        break
                    else:
                        current = current.right

    def collect_data(self):
        data = []
        self._collect_data(self.root, data)
        return data
    
    def _collect_data(self, current, data):
        if current is not None:
            self._collect_data(current.left, data)
            data.append(current)
            self._collect_data(current.right, data)

    def bubble_sort_data(self, data):
        n = len(data)
        for i in range(n):
            for j in range(0, n-i-1):
                if data[j].usia > data[j+1].usia:
                    data[j], data[j+1] = data[j+1], data[j]


def menu_utama():
    print("===================== MENU UTAMA ========================")
    print("1. Tampilkan seluruh data antropometri berdasarkan usia")
    print("2. cari data antropometri berdasarkan tinggi badan")
    print("3. cari data antropometri berdasarkan berat badan")
    print("=========================================================")


def main():
    bst = BST()
    bst.insert(51, 'Monsieur', 21, 175, 66)
    bst.insert(32, 'Allan', 25, 188, 74)
    bst.insert(74, 'Oliver', 17, 160, 59)
    bst.insert(23, 'Andre', 19, 173, 69)
    bst.insert(42, 'Cooper', 23, 190, 73)
    bst.insert(67, 'Alex', 20, 157, 57)
    bst.insert(82, 'Leonardo', 24, 162, 56)
    bst.insert(11, 'James', 18, 173, 63)
    bst.insert(35, 'Robert', 22, 185, 72)
    bst.insert(45, 'Evan', 26, 168, 65)
    while True:
        menu_utama()
        pilihan = int(input("Masukkan pilihan: "))
        if pilihan == 1:
            data = bst.collect_data()
            bst.bubble_sort_data(data)
            print("Data Antropometri:")
            for node in data:
                print(f"ID: {node.id}, Nama: {node.nama}, Usia: {node.usia}, Tinggi Badan: {node.tinggi_badan}, Berat Badan: {node.berat_badan}")
        
        elif pilihan == 2:
            found = False
            while found == False:
                tb = int(input("Masukkan tinggi badan: "))
                data = bst.collect_data()
                print("Data Antropometri Berdasarkan Tinggi Badan:")
                for node in data:
                    if node.tinggi_badan == tb:
                        print(f"ID: {node.id}, Nama: {node.nama}, Usia: {node.usia}, Tinggi Badan: {node.tinggi_badan}, Berat Badan: {node.berat_badan}")
                        found = True
                if not found:
                    print("Data tidak ditemukan, silahkan input kembali")
                    found = False

        elif pilihan == 3:
            found = False
            while found == False:
                bb = int(input("Masukkan tinggi badan: "))
                data = bst.collect_data()
                print("Data Antropometri Berdasarkan Berat Badan:")
                for node in data:
                    if node.berat_badan == bb:
                        print(f"ID: {node.id}, Nama: {node.nama}, Usia: {node.usia}, Tinggi Badan: {node.tinggi_badan}, Berat Badan: {node.berat_badan}")
                        found = True
                if not found:
                    print("Data tidak ditemukan, silahkan input kembali")
                    found = False
        else:
            print("Pilihan tidak valid, silahkan input kembali")
            continue

main()
```
