### sum of all the node in the generic Tree 

```python
import sys
import queue

class TreeNode :
    def __init__(self, data) :
        self.data = data
        self.children = list()

def inputLevelWise(li) :
    i = 0
    data = li[i] 
    i += 1
    if data == -1 :
        return None
    root = TreeNode(data) 
    q = queue.Queue()
    q.put(root)
    while (not q.empty()) :
        frontNode = q.get()
        noOfChildren = li[i]
        i += 1
        childrenArray = li[i : i+noOfChildren]
        for childData in childrenArray :
            childNode = TreeNode(childData)
            frontNode.children.append(childNode)
            q.put(childNode)
        i = i+noOfChildren
    return root
        
def sumOfAllNodes(root) :
    if root is None :
        return 
    sumNode = root.data 
    for child in root.children :
        sumNode += sumOfAllNodes(child)
    return sumNode 
#main
sys.setrecursionlimit(10**6)
li = [int(elem) for elem in list(input().strip().split())]
root = inputLevelWise(li)
print(sumOfAllNodes(root))
```

```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data = data 
        self.left = None 
        self.right = None 

class BST :
    def __init__(self):
        self.root = None 
        self.NumCount = 0 

    @staticmethod
    def insertHelper(root,data) :
        if root is None :
            node = BinaryTreeNode(data)
            return node
        if root.data >= data :
            root.left = BST.insertHelper(root.left,data)
            return root 
        else :
            root.right = BST.insertHelper(root.right,data)
            return root 

    def insert(self, data) :
        self.NumCount+=1
        self.root = self.insertHelper(self.root,data)
    @staticmethod
    def minValue(root):
        if root is None :
            return None 
        if root.left is None :
            return root.data
        return BST.minValue(root.left)

    @staticmethod
    def deleteHelper(root,data):
        if root is None :
            return False, None 

        if root.data > data :
            delete,newNode = BST.deleteHelper(root.left,data)
            root.left = newNode
            return delete,root 

        elif root.data < data :
            delete,newNode = BST.deleteHelper(root.right,data)
            root.right = newNode
            return delete, root 

        elif root.left == None and root.right == None :
            return True , None 

        elif root.left is None :
            return True, root.right 

        elif root.right is None :
            return True , root.left 

        else :
            replacement = BST.minValue(root.right) 
            root.data  = replacement
            delete, newNode = BST.deleteHelper(root.right, replacement)
            root.right = newNode
            return delete , root 
    def delete(self, data):
        remove, node = BST.deleteHelper(self.root,data)
        if remove :
            self.NumCount-=1 
        self.root = node 
        return remove

    @staticmethod
    def isPresentHealper(root,data):
        if root is None :
            return False
        if root.data == data :
            return True 
        if root.data > data :
            return BST.isPresentHealper(root.left,data)
        else :
            return BST.isPresentHealper(root.right,data)

    def isPresent(self, data):
        return self.isPresentHealper(self.root,data)
    
    def count(self):
        return self.NumCount
    
    @staticmethod
    def printTreeHelper(root):
        if root is None :
            return None 
        print(root.data,end = ':')
        if root.left != None :
            print("L:{}".format(root.left.data),end = ',')
        if root.right != None :
            print("R:{}".format(root.right.data),end = '')
        print()
        BST.printTreeHelper(root.left)
        BST.printTreeHelper(root.right)  

    def printTree(self):
        self.printTreeHelper(self.root)
        
b = BST()
q = int(input())
while (q > 0) :
    li = [int(ele) for ele in input().strip().split()]
    choice = li[0]
    q-=1
    if choice == 1:
        data = li[1]
        b.insert(data)
    elif choice == 2:
        data = li[1]
        b.delete(data)
    elif choice == 3:
        data = li[1]
        ans = b.isPresent(data)
        if ans is True:
            print('true')
        else:
            print('false')
    else:
        b.printTree()
```
