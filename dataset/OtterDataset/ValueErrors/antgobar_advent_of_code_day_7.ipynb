```python
with open("day_7_input.txt", "r") as f:
    raw_data = f.readlines()

data = [d.strip("\n") for d in raw_data]

events = []

for event in data:
    if "ls" in event:
        continue
    if "cd" in event:
        args = event.split(" ")
        events.append((True, args[-1]))
        continue
    events.append((False, int(args[0])))
```

```python
def keep_useful_info(entry):
    # keep only directory changes and file size
    return (True, entry[1]) if entry[0] == "$ cd" else (False, int(entry[0])) if entry[0][0].isdigit() else None
```

```python
list(filter(bool, map(keep_useful_info, (line.strip('\n').rsplit(' ', 1) for line in data))))
```
