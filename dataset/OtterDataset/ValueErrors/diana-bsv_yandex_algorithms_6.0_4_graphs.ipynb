```python
#D
class Leaf:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        
def print_(root, d):
    if not root:
        return
    print_(root.left, d + 1)
    print("." * d + str(root.val))
    print_(root.right, d + 1)
    
tree = None

f = open("input.txt", "r")

for line in f.readlines():
    if line[:9] == "PRINTTREE":
        print_(tree, 0)
        
    elif line[:3] == "ADD":
        n = int(line.split()[1])
        if not tree:
            tree = Leaf(n)
            print("DONE")
        else:
            tmp = tree
            p = tree
            while(tmp):
                
                if tmp.val == n:
                    print("ALREADY")
                    break
                    
                elif tmp.val > n:
                    p = tmp
                    tmp = tmp.left
                else:
                    p = tmp
                    tmp = tmp.right
                    
            if not tmp:
                if p.val > n:
                    p.left = Leaf(n)
                else:
                    p.right = Leaf(n)
                print("DONE")
    else:
        n = int(line.split()[1])
        if not tree:
            print("NO")
        else:
            tmp = tree
            while(tmp):
                
                if tmp.val == n:
                    print("YES")
                    break
                    
                elif tmp.val > n:
                    tmp = tmp.left
                else:
                    tmp = tmp.right
                    
            if not tmp:
                print("NO")
        
        

f.close()
```

```python
MOD = 10**9 + 7

from collections import defaultdict
import sys
sys.setrecursionlimit(3000)

# Факториалы и обратные факториалы для комбинаций
def precompute_factorials(n, mod):
    fact = [1] * (n + 1)
    inv_fact = [1] * (n + 1)
    for i in range(2, n + 1):
        fact[i] = fact[i - 1] * i % mod
    inv_fact[n] = pow(fact[n], mod - 2, mod)  # Обратный факториал последнего элемента
    for i in range(n - 1, 0, -1):
        inv_fact[i] = inv_fact[i + 1] * (i + 1) % mod
    return fact, inv_fact

def comb(n, k, fact, inv_fact, mod):
    if n < k or k < 0:
        return 0
    return fact[n] * inv_fact[k] % mod * inv_fact[n - k] % mod

def dfs(v, parent, tree, dp, size, fact, inv_fact, mod):
    dp[v] = 1
    size[v] = 1
    for u in tree[v]:
        if u != parent:
            dfs(u, v, tree, dp, size, fact, inv_fact, mod)
            dp[v] = dp[v] * dp[u] % mod
            dp[v] = dp[v] * comb(size[v] + size[u] - 1, size[u], fact, inv_fact, mod) % mod
            size[v] += size[u]

def count_topological_sorts(n, edges):
    tree = defaultdict(list)
    for a, b in edges:
        tree[a].append(b)
        tree[b].append(a)
    
    fact, inv_fact = precompute_factorials(n, MOD)
    dp = [0] * (n + 1)
    size = [0] * (n + 1)
    
    dfs(1, -1, tree, dp, size, fact, inv_fact, MOD)
    return dp[1]

# Чтение данных
n = int(input())
edges = [tuple(map(int, input().split())) for _ in range(n - 1)]

# Вывод результата
print(count_topological_sorts(n, edges))
```

```python
from collections import defaultdict

def find_max_product(N, edges):
    # Построение графа
    graph = defaultdict(list)
    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)

    # DFS для нахождения самого удаленного узла и длины диаметра
    def dfs(node, parent, depth):
        farthest_node = node
        max_depth = depth
        for neighbor in graph[node]:
            if neighbor != parent:
                new_node, new_depth = dfs(neighbor, node, depth + 1)
                if new_depth > max_depth:
                    farthest_node, max_depth = new_node, new_depth
        return farthest_node, max_depth

    # Найти диаметр дерева
    first_end, _ = dfs(1, -1, 0)
    second_end, diameter = dfs(first_end, -1, 0)

    # DFS для вычисления размеров поддеревьев и максимальных путей
    subtree_sizes = [0] * (N + 1)
    max_path = [0] * (N + 1)
    second_max_path = [0] * (N + 1)

    def calculate_subtrees_and_paths(node, parent):
        subtree_sizes[node] = 1
        for neighbor in graph[node]:
            if neighbor != parent:
                calculate_subtrees_and_paths(neighbor, node)
                subtree_sizes[node] += subtree_sizes[neighbor]
                path_length = max_path[neighbor] + 1
                if path_length > max_path[node]:
                    second_max_path[node] = max_path[node]
                    max_path[node] = path_length
                elif path_length > second_max_path[node]:
                    second_max_path[node] = path_length

    calculate_subtrees_and_paths(1, -1)

    # Найти максимальное произведение
    max_product = 0
    for node in range(1, N + 1):
        if second_max_path[node] > 0:
            max_product = max(max_product, max_path[node] * second_max_path[node])

    return max_product

# Пример использования
N = 4
edges = [
    (1, 2),
    (2, 3),
    (3, 4)
]
print(find_max_product(N, edges))
```

```python
MOD = 10**9 + 7

def solve_tree_toposort(N, edges):
    from math import comb
    from collections import defaultdict
    import sys
    sys.setrecursionlimit(3000)

    # Список смежности
    graph = defaultdict(list)
    for a, b in edges:
        graph[a].append(b)

    # Предвычисление факториалов и обратных элементов
    fact = [1] * (N + 1)
    inv_fact = [1] * (N + 1)
    for i in range(2, N + 1):
        fact[i] = fact[i - 1] * i % MOD
    inv_fact[N] = pow(fact[N], MOD - 2, MOD)
    for i in range(N - 1, 0, -1):
        inv_fact[i] = inv_fact[i + 1] * (i + 1) % MOD

    # Функция для подсчета размера поддерева и DP
    dp = [0] * (N + 1)
    size = [0] * (N + 1)

    def dfs(node):
        size[node] = 1
        dp[node] = 1
        child_sizes = []
        for child in graph[node]:
            dfs(child)
            size[node] += size[child]
            dp[node] = dp[node] * dp[child] % MOD
            child_sizes.append(size[child])

        # Количество перестановок поддеревьев
        total = size[node] - 1
        for cs in child_sizes:
            dp[node] = dp[node] * comb(total, cs) % MOD
            total -= cs

    # Запуск DFS из корня
    dfs(1)
    return dp[1]

# Ввод
N = int(input())
edges = []
for _ in range(N - 1):
    a, b = map(int, input().split())
    edges.append((a, b))

# Вывод
print(solve_tree_toposort(N, edges))
```
