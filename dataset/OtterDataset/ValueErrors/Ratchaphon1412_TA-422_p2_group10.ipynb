```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload

# from google.colab import drive
# drive.mount("/content/drive", force_remount=True)
```

```python
# ---
# Group No. 10
#1. นางสาวณัฐชา บุญเฉย 6410450435
#2. นางสาวอิงกมล ขาวกมล 6410450656
#3. นางสาวชลดา บำรุงรส 6410450401
#4. นายณรงค์เดช สำเร็จประสงค์ 6410450427
# ---

group = 10 # กลุ่มที่

problem = 2 # ปัญหาที่
data_num = 1 # ชุดข้อมูลที่

# ชุดข้อมูลเก็บใน /in
in_path = "./dataset/P"+str(problem)+"/in/"
data_path = in_path+"p"+str(problem)+"_input_dataset_"+str(data_num)+".txt"

# ผลลัพธ์เก็บใน /out
out_path = "./dataset/P"+str(problem)+"/out/"
result_path = out_path+"p"+str(problem)+"_group"+str(group)+"-output"+str(data_num)+".csv"
png_path = out_path+"p"+str(problem)+"_group"+str(group)+"-plot"+str(data_num)+".png"

# อ่านข้อมูล
data = pd.read_csv(data_path)

# --- เริ่มเขียนโค้ดตรงนี้ ---

data3={'PatientID':[],  'Age':[],   'Gender':[],
             'Diagnosis':[],   'Treatment':[]}

df = pd.DataFrame(data3)

for i in range(len(data)):
  data2 =data.iloc[i,0].split()

  new_row = {'PatientID':data2[0],  'Age':int(data2[1]),   'Gender':data2[2],
             'Diagnosis':data2[3],   'Treatment':data2[4]}
  df.loc[len(df)] = new_row

df = df.groupby(pd.cut(df['Age'], [20, 30, 40, 50, 60])).agg(age_count=('Age','count'))

df = df.sort_values('age_count', ascending=False)

data = df.reset_index()

data= data.rename(columns={"Age": "Age Group","age_count":"Number of Patients"})


data4 = df.sort_values('Age', ascending=True)


plt = reload(plt)
plt.figure(figsize=(10,10))


plt.bar(range(len(data4['age_count'])), data4['age_count'], width=1, edgecolor='black')


plt.xlabel('Age')
plt.ylabel('Number of Patients')

plt.title('Age Distribution of Patients')


# --- สิ้นสุดโค้ดตรงนี้ ---

# เซฟผลลัพธ์
data.to_csv(result_path, index=False)

# แสดงผลลัพธ์
data.head()

# เซฟกราฟ
plt.savefig(png_path, dpi=100)

# แสดงผลกราฟ
plt.show()
```
