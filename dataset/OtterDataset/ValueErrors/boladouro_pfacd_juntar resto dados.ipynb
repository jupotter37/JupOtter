```python
from pyprojroot.here import here
import pandas as pd
from pyarrow import feather

# import here('data/AllDaysEventos_com_zonas.feather')
dias = feather.read_feather(here('data/AllDaysEventos_com_zonas.feather'))
dias
```

```python
# import here('data/QuebrasJuntoInc_com_zonas.xlsx')
eventos = pd.read_excel(here('data/QuebrasJuntoInc_com_zonas.xlsx'))
eventos
```

```python
# uniform concelhos
from unidecode import unidecode
def transform_concelhos(x):
    return unidecode(x).upper().replace(' ', '-')

eventos['Concelho'] = eventos['Concelho'].apply(transform_concelhos)
dias['Concelho'] = dias['Concelho'].apply(transform_concelhos)
eventos
```

```python
# import here('data/concelho_degress.csv')
degress = pd.read_csv(here('data/concelho_degress.csv'))
# import here('data/subestacoes AT_MT por concelho.csv')
subestacoes = pd.read_csv(here('data/subestacoes AT_MT por concelho.csv'))
# left join por concelho
degress['concelho'] = degress['concelho'].apply(transform_concelhos)
subestacoes['concelho'] = subestacoes['concelho'].apply(transform_concelhos)
eventos = eventos.merge(degress, left_on='Concelho', right_on='concelho', how='left')
eventos = eventos.merge(subestacoes, left_on='Concelho', right_on='concelho', how='left')
dias = dias.merge(degress, left_on='Concelho', right_on='concelho', how='left')
dias = dias.merge(subestacoes, left_on='Concelho', right_on='concelho', how='left')
eventos
```

```python
# change column names: 
# total_concelho -> total_concelho_zonas
# degree -> soma_degree_rede_concelho
# count -> n_subestacoes_concelho
eventos.rename(columns={'total_concelho': 'total_concelho_zonas', 'degree': 'soma_degree_rede_concelho', 'count': 'n_subestacoes_concelho'}, inplace=True)
dias.rename(columns={'total_concelho': 'total_concelho_zonas', 'degree': 'soma_degree_rede_concelho', 'count': 'n_subestacoes_concelho'}, inplace=True)
# drop concelho_x, concelho_y
eventos.drop(columns=['concelho_x', 'concelho_y'], inplace=True)
dias.drop(columns=['concelho_x', 'concelho_y'], inplace=True)
eventos
```

```python
# where soma_degree_rede_concelho is null, fill with 0 (same thing with n_subestacoes_concelho)
eventos['soma_degree_rede_concelho'].fillna(0, inplace=True)
eventos['n_subestacoes_concelho'].fillna(0, inplace=True)
dias['soma_degree_rede_concelho'].fillna(0, inplace=True)
dias['n_subestacoes_concelho'].fillna(0, inplace=True)
```

```python
eventos['N.º clientes afetados'].apply(type).value_counts()
```

```python
"""
N.º clientes afetados
<class 'str'>    7727
<class 'int'>       1
Name: count, dtype: int64
"""
# diagnose
eventos['N.º clientes afetados'] = eventos['N.º clientes afetados'].apply(lambda x: int(x) if type(x) == str else x)
eventos['N.º clientes afetados'].apply(type).value_counts()
```

```python
# ValueError: invalid literal for int() with base 10: '244 990'
# check line with this
eventos[eventos['N.º clientes afetados'] == '244 990']
# change to 244990
eventos.loc[eventos['N.º clientes afetados'] == '244 990', 'N.º clientes afetados'] = 244990

# now change whole line to int
eventos['N.º clientes afetados'] = eventos['N.º clientes afetados'].astype(int)
```

```python
# In TIEPI MT (min), SAIFI MT (#), SAIDI MT (min), SAIDI MT (min), END (MWh), SAIFI BT (#), SAIDI BT (min): change where ######### to NaN
for col in ['TIEPI MT (min)', 'SAIFI MT (#)', 'SAIDI MT (min)', 'SAIDI MT (min)', 'END (MWh)', 'SAIFI BT (#)', 'SAIDI BT (min)']:
    eventos[col] = eventos[col].apply(lambda x: None if x == '#########' else x)
```

```python
# in eventos, change to float: TIEPI MT (min), SAIFI MT (#), SAIDI MT (min), SAIDI MT (min), END (MWh), SAIFI BT (#), SAIDI BT (min)
# some are already float, and strings have ',' instead of '.'

for col in ['TIEPI MT (min)', 'SAIFI MT (#)', 'SAIDI MT (min)', 'SAIDI MT (min)', 'END (MWh)', 'SAIFI BT (#)', 'SAIDI BT (min)']:
    eventos[col] = eventos[col].apply(lambda x: float(x.replace(',', '.')) if type(x) == str else x)
```

```python
# save both to feather
feather.write_feather(eventos, here('data/Eventos.feather'))
feather.write_feather(dias, here('data/EventosPorDia.feather'))
dias
```
