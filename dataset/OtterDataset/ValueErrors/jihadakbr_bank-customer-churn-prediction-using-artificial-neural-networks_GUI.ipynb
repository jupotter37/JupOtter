```python
import warnings
warnings.filterwarnings("ignore")
from ml_utils import *
```

```python
root = Tk()
logo = Image.open("github.png")
logo = logo.resize((50, 50))
logo = ImageTk.PhotoImage(logo)
logo_label = Label(image=logo)
logo_label.image = logo
logo_label.grid(row=0, column=0, sticky='w', padx=(40,0),pady=(30,0))
Label(root,text = "jihadakbr", font='Courier 16').grid(row=0,column=0,padx=(30,0),pady=(30,0))

root.title('Bank Customer Churn Prediction')
Label(root,text = "Bank Customer Churn Prediction\nUsing Artificial Neural Networks (ANN)", font='Raleway 18 bold').grid(row=1,columnspan=3,pady=(10,40))

canvas = Canvas(root, width=800, height=300)
canvas.grid(columnspan=3,rowspan=9)

def predict_click():
    t1 = CreditScore.get()
    t2 = Geography.get()
    t3 = Gender.get()
    t4 = Age.get()
    t5 = Tenure.get()
    t6 = Balance.get()
    t7 = NumberOfProducts.get()
    t8 = HasCreditCard.get()
    t9 = IsActiveMember.get()    
    t10 = EstimatedSalary.get()
    
    try:
        t1 = float(t1)
        t4 = float(t4)
        t5 = float(t5)
        t6 = float(t6)
        t7 = float(t7)
        t10 = float(t10)
        
        if t1 != int(t1):
            raise ValueError('Error in "Credit Score": Enter an integer number')
        elif t1 < 0:
            raise ValueError('Error in "Credit Score": Enter a positive number')
        elif t1 > 1000:
            raise ValueError('Error in "Credit Score": Enter a number ≤ 1000')
            
        if t4 != int(t4):
            raise ValueError('Error in "Age": Enter an integer number')    
        elif t4 < 0:
            raise ValueError('Error in "Age": Enter a positive number')
        elif t4 > 100:
            raise ValueError('Error in "Age": Enter a number ≤ 100')
            
        if t5 != int(t5):
            raise ValueError('Error in "Tenure": Enter an integer number')      
        elif t5 < 0:
            raise ValueError('Error in "Tenure": Enter a positive number')
        elif t5 > 10:
            raise ValueError('Error in "Tenure": Enter a number ≤ 10')
            
        if t6 < 0:
            raise ValueError('Error in "Balance": Enter a positive number')
        
        if t7 != int(t7):
            raise ValueError('Error in "Number Of Products": Enter an integer number')   
        elif t7 < 0:
            raise ValueError('Error in "Number Of Products": Enter a positive number')
        elif t7 > 4:
            raise ValueError('Error in "Number Of Products": Enter a number ≤ 4')
            
        if t10 < 0:
            raise ValueError('Error in "Estimated Salary": Enter a positive number')    
        
        if  t2 == "select one":
            messagebox.showerror('Error', 'Error in "Geography": Select an option')
        elif t3 == "select one":
            messagebox.showerror('Error', 'Error in "Gender": Select an option')
        elif t8 == "select one":
            messagebox.showerror('Error', 'Error in "Has Credit Card?": Select an option')
        elif t9 == "select one":
            messagebox.showerror('Error', 'Error in "Is Active Member?": Select an option')        
        else:    
            if Geo_com.get() == "France":
                Germany = 0
                Spain = 0
            elif Geo_com.get() == "Germany":
                Germany = 1
                Spain = 0
            else:
                Germany = 0
                Spain = 1

            if Gen_com.get() == "Male":
                Male = 1
            else:
                Male = 0

            if HCC_com.get() == "Yes":
                Has_Credit_Card = 1
            else:
                Has_Credit_Card = 0 

            if IAM_com.get() == "Yes":
                Is_Active_Member = 1
            else:
                Is_Active_Member = 0 

    except ValueError as ve:
        if "could not convert string to float" in str(ve):
            messagebox.showerror('Error', 'Please fill out all fields\nand enter a valid number')
        elif "invalid literal for int()" in str(ve):
            messagebox.showerror('Error', 'Select an option')
        else:
            messagebox.showerror('Error', str(ve))
        
    p1 = float(t1) # CreditScore
    p2 = float(t4) # Age
    p3 = float(t5) # Tenure
    p4 = float(t6) # Balance
    p5 = float(t7) # NumOfProducts
    p6 = float(Has_Credit_Card) # HasCrCard
    p7 = float(Is_Active_Member) # IsActiveMember
    p8 = float(t10) # EstimatedSalary
    p9 = float(Germany) # Geography_Germany
    p10 = float(Spain) # Geography_Spain
    p11 = float(Male) # Gender_Male
    
    DeepNN = models.load_model("DeepNN.h5", custom_objects={'F1': F1})
    scaler = joblib.load('scalerDeepNN.joblib')
    
    input_data = np.array([[p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11]])
    
    std_data = scaler.transform(input_data)
    prediction = DeepNN.predict(std_data)
    prediction_label = np.round(prediction)

    if(prediction_label[0] == 0):
        result.configure(text="Not Exited", bg="yellow", fg="red")
    else:   
        result.configure(text="Exited", bg="yellow", fg="red")

           
def click1(*args):
    e1.delete(0, 'end')
def click4(*args):
    e4.delete(0, 'end')
def click5(*args):
    e5.delete(0, 'end')    
def click6(*args):
    e6.delete(0, 'end')
def click7(*args):
    e7.delete(0, 'end')
def click10(*args):
    e10.delete(0, 'end')  


#### String Var ###    
CreditScore = StringVar(root)
Age = StringVar(root)
Tenure = StringVar(root)
Balance = StringVar(root)
NumberOfProducts = StringVar(root)
EstimatedSalary = StringVar(root)


########## column 0 ##########
        
# CreditScore, Geography, Gender, Age
e1 = Entry(root, width=40, textvariable=CreditScore)
e4 = Entry(root, width=40, textvariable=Age)

### CreditScore ###
e1.insert(0, 'type here ...')
e1.bind("<Button-1>", click1)

Label(root, text='Credit Score').grid(row=2,column=0, sticky='w',padx=(40,0),pady=(0,5))
e1.grid(row=3,column=0, sticky='w',padx=(40,0),ipady=10,pady=(0,25))

### Geography ###
Geography = StringVar(root)
Geography.set("select one")
Geo_com = ttk.Combobox(root, textvariable=Geography, value=['France','Germany','Spain'], width=37, state="readonly")
Geo_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Geography').grid(row=4,column=0, sticky='w',padx=(40,0),pady=(0,5))
Geo_com.grid(row=5,column=0, sticky='w',padx=(40,0),ipady=10,pady=(0,25))

### Gender ###
Gender = StringVar(root)
Gender.set("select one")
Gen_com = ttk.Combobox(root, textvariable=Gender, value=['Male','Female'], width=37, state="readonly")
Gen_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Gender').grid(row=6,column=0, sticky='w',padx=(40,0),pady=(0,5))
Gen_com.grid(row=7,column=0, sticky='w',padx=(40,0),ipady=10,pady=(0,25))

### Age ###
e4.insert(0, 'type here ...')
e4.bind("<Button-1>", click4)

Label(root, text='Age (years)').grid(row=8,column=0, sticky='w',padx=(40,0),pady=(0,5))
e4.grid(row=9,column=0, sticky='w',padx=(40,0),ipady=10,pady=(0,50))


########## column 1 ##########

# Tenure, Balance, NumberOfProducts, HasCreditCard
e5 = Entry(root, width=40, textvariable=Tenure)
e6 = Entry(root, width=40, textvariable=Balance)
e7 = Entry(root, width=40, textvariable=NumberOfProducts)

### Tenure ###
e5.insert(0, 'type here ...')
e5.bind("<Button-1>", click5)

Label(root, text='Tenure (years)').grid(row=2,column=1, sticky='w',padx=(40,0),pady=(0,5))
e5.grid(row=3,column=1,padx=40,ipady=10,pady=(0,25))

### Balance ###
e6.insert(0, 'type here ...')
e6.bind("<Button-1>", click6)

Label(root, text='Balance ($)').grid(row=4,column=1, sticky='w',padx=(40,0),pady=(0,5))
e6.grid(row=5,column=1,padx=40,ipady=10,pady=(0,25))

### NumberOfProducts ###
e7.insert(0, 'type here ...')
e7.bind("<Button-1>", click7)

Label(root, text='Number Of Products').grid(row=6,column=1, sticky='w',padx=(40,0),pady=(0,5))
e7.grid(row=7,column=1,padx=40,ipady=10,pady=(0,25))

### HasCreditCard ###
HasCreditCard = StringVar(root)
HasCreditCard.set("select one")
HCC_com = ttk.Combobox(root, textvariable=HasCreditCard, value=['Yes','No'], width=37, state="readonly")
HCC_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Has Credit Card?').grid(row=8,column=1, sticky='w',padx=(40,0),pady=(0,5))
HCC_com.grid(row=9,column=1,padx=40,ipady=10,pady=(0,50))


########## column 2 ##########

# IsActiveMember, EstimatedSalary
e10 = Entry(root, width=40, textvariable=EstimatedSalary)

### IsActiveMember ###
IsActiveMember = StringVar(root)
IsActiveMember.set("select one")
IAM_com = ttk.Combobox(root, textvariable=IsActiveMember, value=['Yes','No'], width=37, state="readonly")
IAM_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Is Active Member?').grid(row=2,column=2, sticky='w',padx=(0,40),pady=(0,5))
IAM_com.grid(row=3,column=2, sticky='e',padx=(0,40),ipady=10,pady=(0,25))

### EstimatedSalary ###
e10.insert(0, 'type here ...')
e10.bind("<Button-1>", click10)

Label(root, text='Estimated Salary ($)').grid(row=4,column=2, sticky='w',padx=(0,40),pady=(0,5))
e10.grid(row=5,column=2, sticky='e',padx=(0,40),ipady=10,pady=(0,25))


########## Prediction Button ##########
Label(root, text='').grid(row=6,column=2, sticky='w',padx=(0,40),pady=(0,5))
button = Button(root,text="Predict",command=predict_click,bg="#20bebe", fg="white", font="Raleway 10 bold",
       height=2, width=12)
button.grid(row=7,column=2,sticky='w',pady=(0,25))

#Label(root, text='').grid(row=11,column=8, sticky='w',padx=(0,40),pady=(0,5))
#Label(root, text='').grid(row=12,column=9, sticky='w',padx=(0,40),ipady=10,pady=(0,25))

########## Result ##########
result = Label(root, text="Result",relief="sunken", fg="#666666", bg="#cccccc",font="Raleway 10",height=2, width=12)
result.grid(row=7, column=2,ipady=4,sticky='e',padx=(0,40),pady=(0,25))

root.mainloop()
```
