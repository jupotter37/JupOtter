```python
#compute each doc vector then compare (faster)
def compute_doc_vec(docs):
    num_not_found=0

    docs_vecs={}
    for doc in docs:
        doc_vectors=[]
        for word in doc[2].split():
            try:
                doc_vectors.append(w2v[re.sub(r'\W+', '',word)])
            except: 
                try:
                    word=word.lower()
                    doc_vectors.append(w2v[re.sub(r'\W+', '',word)])
                except:
                    if num_not_found <1000:
                        print(word)
                    num_not_found+=1
        # find avg_vector by column-wise sum
        avg_vec=  np.sum(doc_vectors, axis=0)/ len(doc_vectors)
        docs_vecs[doc[0]] = avg_vec  
            
            
    print("len docs_vecs",len(docs_vecs)," should be 81359 ")   
    print('num not found', num_not_found)

    return docs_vecs



docs_vecs= compute_doc_vec(headline_text) 
```

```python
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances

def compute_avg_vector(words, w2v):
    v=[]
    for word in words.split():
        try:
            v.append(w2v[word])
        except: 
            pass #print('not found', word)
    v=  np.sum(np.array(v), axis=0)/ len(v)
    return v

    
    
    
def get_doc(k, docs):
    for key, label, doc in docs:
        if key == k: return doc
    


def find_sim_doc(vec, docs_vecs,docs):
    best_sim=100
    best_doc=""
    best_key=0
#    print("vec", len(vec), "doc vec", len(docs_vecs))
    
    for key, doc_vec in docs_vecs.items():
        
        aa = vec.reshape(1,-1)
        ba = doc_vec.reshape(1,-1)
        sim = cosine_distances(aa, ba)
        if sim < best_sim:
            best_sim = sim
            best_doc = get_doc(key, docs)
            best_key=key
            #print('found better doc', best_sim)
    return best_doc, best_sim, best_key

def find_sim_doc_using_scipy(vec, docs_vecs_list, docs):
    



    aa = vec.reshape(1,-1) # (1,300)
    ba = docs_vecs_list
    
    r=scipy.spatial.distance.cdist(aa,ba ,'cosine')
   # print(len(r[0]))
    idx = np.argmin(r)
    return docs[idx] , r[0][idx], idx

```

```python
# compute tfidf of headline_doc
from sklearn.feature_extraction.text import CountVectorizer
import re
from nltk.corpus import stopwords
stopwords.words('english')

stopwords_list= stopwords.words('english')
stopwords_list.append('the')

#create a vocabulary of words, 
#ignore words that appear in 85% of documents, 
#eliminate stop words
cv=CountVectorizer(max_df=0.85,stop_words=stopwords_list, max_features=10000)
word_count_vector=cv.fit_transform(headline_docs)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count_vector)
```

```python
def sort_coo(coo_matrix):    
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

def extract_topn_from_vector(feature_names, sorted_items, topn=10):
    """get the feature names and tf-idf score of top n items"""
    
    #use only topn items from vector
    sorted_items = sorted_items[:topn]

    score_vals = []
    feature_vals = []

    for idx, score in sorted_items:
        fname = feature_names[idx]
        
        #keep track of feature name and its corresponding score
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    #create a tuples of feature,score
    #results = zip(feature_vals,score_vals)
    results= {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]]=score_vals[idx]
    
    return results
```

```python
# you only needs to do this once
feature_names=cv.get_feature_names()
len(feature_names)
```

```python
def find_sim_docs_using_scipy(vec, docs_vecs_list, docs, n):
    '''
    find the most similar n docs
    '''
    aa = vec.reshape(1,-1) # (1,300)
    ba = docs_vecs_list
    
    r=scipy.spatial.distance.cdist(aa,ba ,'cosine')

    # keep in mind this (argpartition and partition) does not return result ordered 
    temp = np.argpartition(r[0], n)
    result_args = temp[:n]
    temp = np.partition(r[0], n)
    result = temp[:n] 
    # return documents , sim_value, indies
    sim_docs=[]
    for ind in result_args:
        sim_docs.append(docs[ind][2])


    return sim_docs , r[0][result_args], result_args
```
