```python
%matplotlib inline
import pytz
import pandas as pd
from datetime import datetime
from zipline.algorithm import TradingAlgorithm
from zipline.utils.factory import load_bars_from_yahoo
from zipline.api import symbol, order, record

# Load data manually from Yahoo! finance
start = pd.Timestamp('2012-01-01', tz='UTC')
end = pd.Timestamp('2015-03-29', tz='UTC')

sp500 = pd.DataFrame.from_csv('SP500.csv')

data = load_bars_from_yahoo(
    stocks=sp500['symbol'].values.tolist()[:10], # first 10 of S&P 500. stocks=['AAPL','A','AA'],
    start=start, end=end)
```

```python
from zipline.api import order_target
# Define algorithm
def initialize(context):
    pass
    
def handle_data(context, data):
    for stk in data:
        stock = data[stk]
        if not pd.isnull(stock) :
            ## order_target(symbol(stk), 10)
            record(stk=stock.price)

# Create algorithm object passing in initialize and
# handle_data functions
algo_obj = TradingAlgorithm(initialize=initialize, 
                            handle_data=handle_data)

# Run algorithm
perf = algo_obj.run(data)
```
