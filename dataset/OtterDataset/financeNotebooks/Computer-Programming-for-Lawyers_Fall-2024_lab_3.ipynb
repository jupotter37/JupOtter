```python
# Reviewing time_elapsed (20 minutes)

MINUTES_IN_HOUR = 60
MINUTES_IN_DAY = 1440
MINUTES_IN_HALFDAY = 720

# Recap of Assignment Specifications.
# Input: # of minutes elapsed from midnight (if 12:01am -> 1)
# Output: What the clock would say (if input == 1 -> 12:01 am)
print("start time: 12:00 am")
minutes_input = input("minuites since midnight please:")
minutes_input_int = int(minutes_input)
print(minutes_input_int)

# HH (HOURS)
hours_output = minutes_input_int % MINUTES_IN_DAY // MINUTES_IN_HOUR
print("hours output:"+str(hours_output))


# MM (MINUTES) -> DOUBLE DIGITS
# Modulo for time left after the hours
minutes_output = minutes_input_int % MINUTES_IN_HOUR
print(minutes_output)

# am/pm tracking




```

```python
# For Loops! (10 minutes)

# Importing text files

# One thing to do before writing a single line of code is...

filename = input("Filename: ")

# File has one word per line. Split words into list.
words = open(filename).read().split('\n')
badger = input('find this word:')
counter = 0

for word in words:
    # Print each word.
    counter = counter + 1
    
    print (str(counter)+' '+word)
    if word == badger:
        print('match for '+badger+' at ' + str(counter))
        break
    

print('wow we are done!')


# 1 the
# 2 of
# 3 and
```
