```python
import json
import pandas as pd
```

```python
# Load questions
with open('../data/datasets/new_eval_dataset.json') as f:
    questions = json.loads(f.read())
```

```python
# Load documents
docs = json.load(open('all_data_collected.json'))
docs_df = pd.DataFrame(docs, columns=['id', 'text'])
```

```python
# Example once the span and doc is found
def set_doc_and_span(number, doc_and_spans):
    evaluations = []
    for doc_and_span in doc_and_spans:
        ev = {}
        ev['doc'] = doc_and_span[0]
        ev['spans'] = [{'char_end': '', 'char_start': '', 'response': doc_and_span[1]}]
        evaluations.append(ev)
    questions[number]['eval'] = evaluations
```

```python
def get_best_documents(keywords):
    f = docs_df['text'].str.contains(keywords[0], case=False, regex=False)
    for keyword in keywords[1:]:
        f = f & docs_df['text'].str.contains(keyword, case=False, regex=False)
    return docs_df[f]
```

```python
def display_cfa_text(text):
    for t in text.split('\n'):
        if t == '':
            continue
        print(t)
        print()
```

```python
def get_text_by_id(docs_df, doc_id):
    index = docs_df[docs_df['id'] == doc_id].index
    return docs_df.iloc[index[0]]['text']
```
