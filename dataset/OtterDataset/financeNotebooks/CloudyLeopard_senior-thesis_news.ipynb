# News API
Used to be Google News API

https://newsapi.org/

Documentation: https://newsapi.org/docs/get-started

Note: you can also filter by news source/publisher

```python
import os
from dotenv import load_dotenv
import requests

# load environment variables
load_dotenv()

NEWS_API = os.getenv("NEWS_API_KEY")
```

## Search for Article
https://newsapi.org/docs/get-started#search

https://newsapi.org/docs/endpoints/everything

```python
params = {
    "apiKey": NEWS_API,
    "q": "Microsoft",
    "sources": "the-wall-street-journal"
}

r = requests.get("https://newsapi.org/v2/everything", params=params)
res = r.json()
res
```

## Top Headline
https://newsapi.org/docs/endpoints/top-headlines

```python
params = {
    "category": "business",
    "apiKey": NEWS_API,
    "country": "us"
}

r = requests.get("https://newsapi.org/v2/top-headlines", params=params)
res = r.json()
res
```

## Get Sources
https://newsapi.org/docs/endpoints/sources

```python
params = {
    "apiKey": NEWS_API,
    "category": "business",
    "country": "us"
}
r = requests.get("https://newsapi.org/v2/top-headlines/sources", params=params)
res = r.json()
res
```
