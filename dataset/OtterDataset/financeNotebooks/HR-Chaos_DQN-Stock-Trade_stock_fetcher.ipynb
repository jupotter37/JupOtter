```python
import yfinance as yf
import requests
import pandas as pd

# Alpha vantage api key: 
```

```python
# test data for google stock price
ticker_symbol = 'AAPL'
ticker_data = yf.Ticker(ticker_symbol)
historical_data = ticker_data.history(period='1d', start='2010-1-1')
print(historical_data.shape)
print(historical_data)
```

```python
# load historical_data into a csv file
pd.DataFrame(historical_data).to_csv('AAPL_stock.csv', index=False)
```

```python
# test alpha vantage api

API_URL = "https://www.alphavantage.co/query"
API_KEY = ""

parameters = {
    "function": "TIME_SERIES_DAILY",
    "symbol": "MSFT",
    "apikey": API_KEY
}

response = requests.get(API_URL, params=parameters)

data = response.json()
```

```python
import json

pretty_data = json.dumps(data, indent=4)
print(pretty_data)
```

```python
# check MSFT from alpha vantage
parameters = {
    "function": "NEWS_SENTIMENT",
    "tickers": "MSFT",
    "apikey": API_KEY
}
r = requests.get(API_URL, params=parameters)
data = r.json()

pretty_data = json.dumps(data, indent=4)
print(pretty_data)
```
