### Understanding Simple and Compound Interest

Simple Interest is interest earned on a principle, as where compound interest earned on interest.

$FV(simpleInterest) = PV(1 + R * n)$

$FV(compoundInterest) = PV(1 + R)^n$

**Where PV is the present value of the load, R is the rate of interest, and n is the ammount of periods interest is taken.**

Below are an example.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pv = 1000
r = 0.08
n = 10
t = np.linspace(0, n, n)
y1 = np.ones(len(t))*1000
y2 = pv * (1 + r * t)
y3 = pv * (1 + r)**t
plt.figure(figsize=(15,7))
plt.title('Simple Vs. Compound Interest')
plt.xlabel('Number of Years')
plt.ylabel('Values')
plt.xlim(0,11)
plt.ylim(800,2200)
plt.plot(t, y1, 'b-')
plt.plot(t, y2, 'g--')
plt.plot(t, y3, 'r-')
```

### Adding Text to Graphs

```python
x = [0, 1, 2]
y = [2, 4, 6]

plt.figure(figsize=(15,7))
plt.plot(x,y)
plt.figtext(0.2, 0.7, 'North & West')
plt.figtext(0.7, 0.2, 'South & East')
```
