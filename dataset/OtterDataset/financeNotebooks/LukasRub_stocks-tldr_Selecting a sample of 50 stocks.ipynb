Selecting 50 most active stocks from https://www.tradingview.com/markets/stocks-usa/market-movers-active/

```python
datetime.today().strftime("%Y-%m-%d-%H:%M:%S")
```

```python
http = urllib3.PoolManager()
url = "https://www.tradingview.com/markets/stocks-usa/market-movers-active/"

response = http.request('GET', url)
soup = BeautifulSoup(response.data, 'html.parser')
```

```python
table = soup.find("table", {"class": "tv-screener-table"})
trending_stocks = [a.text for a in table.find_all("a") 
                   if a["href"].startswith("/symbols/")]
```

```python
crossref_stocks = stocks[stocks.ticker.isin(trending_stocks)]

display(crossref_stocks)
print(len(crossref_stocks))
```

```python
out_dir = "../../data/processed/available_stocks"
out_file = f"revolut.top50.{datetime.today().strftime('%Y-%m-%dT%H:%M:%S')}.jsonl"


path = Path(out_dir)
path.mkdir(exist_ok=True)
path = path.joinpath(out_file)
```

```python
crossref_stocks.head(50).to_json(path, orient="records", lines=True)
```
