# 5.Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
page2 = requests.get('https://www.cnbc.com/world/?region=world')
page2
```

```python
soup2 = BeautifulSoup(page2.content)
```

```python
Headline = []

for i in soup2.find_all('a',class_="LatestNews-headline"):
    Headline.append(i.text)
    
Headline
```

```python
Time = []

for i in soup2.find_all('span',class_="LatestNews-wrapper"):
    Time.append(i.text)
    
Time
```

```python
News_Link = []

for i in soup2.find_all('a',class_="LatestNews-headline"):
    News_Link.append(i['href'])

News_Link
```

```python
News = pd.DataFrame({'Headline':Headline,'Time':Time,'News_Link':News_Link})
News
```

# 7.Write a python program to scrape mentioned details from and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/welcome-back')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
Name = [] #empty list for store the text

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Name.append(i.text)
```

```python
Location = []

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
    
    
Location
```

```python
Rating = []

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Rating.append(i.text)
    
Rating
```

```python
Cuisine = []

for i in soup.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.replace('₹',''))
    
    
Cuisine
```

```python
Image_url = []

for i in soup.find_all('img',class_="no-img"):
    Image_url.append(i['data-src'])
    
    
Image_url
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Name':Name,'Cuisine':Cuisine,'Location':Location,'Rating':Rating,'Image_URL':Image_url})
df
```
