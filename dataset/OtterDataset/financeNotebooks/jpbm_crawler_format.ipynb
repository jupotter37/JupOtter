```python
import os

DATAPATH = "data/"
DATAFILES = [DATAPATH+file for file 
             in os.listdir(DATAPATH) 
             if '.json' in file]
```

```python
import json

def get_items(paths):
    """
    Generator that yields one article at a time
    """
    for path in paths:
        with open(path,'r') as file:
            for line in file:
                yield json.loads(line)
                
def filter_items(items):
    """
    Generator that filters articles
    
    (In general, it would be better to improve the crawler to
    avoid downloading things that won't be needed)
    """
    
    for item in items:
        
        # check if url is from reuters/finance
        
        url = item['url']
        if 'https://www.reuters.com/finance/' not in url:
            pass
        elif 'sportsNews' in url:
            pass
        else:
            yield item
            
```

```python
d = get_items(DATAFILES)
for i in range(100):
    print(next(d)['url'])
```
