# Using PyTrading for Stock Market Strategy Comparison

PyTrading is a python package designed to facilitate the creation and testing of automated trading strategies. The main objects of the package are portfolios and strategy objects. There are several strategies to choose from in PyTrading. All of which manage a portfolio.

# Using Portfolios

```python
from Portfolio import Portfolio

# first lets create a portfolio
p = Portfolio(10000)

def print_status(p):
    print("Equity is: {}".format(p.equity.items()))
    print("Balance: {}".format(p.balance))
    
# lets buy a stock
p.buy('AAPL',price=10.0,shares=7)
print_status(p)
# The portfolio automatically handles comission (percentage and/or flat rate)

# And knows how much you can afford to buy
p.buy_max('VZ',price=2.56)
print_status(p)

# Will automatically sell all stocks and adjust for comission/flat rate
p.sell_max('VZ',price=20)
print_status(p)

# Can buy weighted linear combos
# Good for designing optimal portfolios
p.batch_buy(['AAPL','AA','VZ'],prices=[1,34,2],weights=[.25,.25,.5])
print_status(p)
```

These portfolio objects make it easy to manage one's finances and are the core building block of the strategy object. Each strategy object manages a portfolio. The strategy will observe the day's stock data, and then act on it by buying and selling stocks through the portfolio.

# Using Strategies

The first strategy we will encounter is the buy and hold strategy. It takes in a starting balance, and a list of tickers and will buy equal proportions of the stocks in the tickers on the first day of trading. From this point on the strategy will not sell any of the assets There are several parameters to change the behavior of this strategy inclding weights to priotitize the buying of particular stocks in the portfolio. Strategies can be backtested using the backtest function which returns the percentage gains achieved in equity. One can toggle the starting and ending dates of the backtest function and whether or not the value of equity is calculated before commision with the correct=False flag .

```python
from Strategies import BuyAndHoldStrategy, backtest

tickers = ['AAPL','VZ']
bs = BuyAndHoldStrategy(10000, tickers)
percent_gains = backtest(bs,start="2014-1-1", end="2015-11-02",correct=False)
print('Buy and hold strategy made a %{} return on investment'.format(round(percent_gains,1)) )
```

PyTrading also has a momentum strategy which sells a stock if its price falls below its moving average and buys a stock if its price rises above its moving average. This strategy accepts a similiar portfolio weighting argument and the number of days in the moving average can be set upon creation with the window parameter.

```python
from Strategies import MomentumStrategy

ms = MomentumStrategy(10000,tickers,window=50)
percent_gains = backtest(ms,start="2014-1-1", end="2015-11-02",correct=False)
print('Momentum strategy made a %{} return on investment'.format(round(percent_gains,1)) )
```

PyTrading Also has an informed buy and hold class which fits a time series estimator to the entire s&p500 and chooses the k best stocks to buy and hold. It must wait a number of days to observe data before it can fit the time series estimator. This strategy can accept any sklearn model (the default is linear regression) and will wait for (default 100) days before it buys and holds the stocks. The envelope is the buy and hold weighting and can be set to uniform, proportional to the projected gains, or proportional to the log of the projected gains.

```python
from Strategies import TSEBuyAndHoldStrategy

ts = TSEBuyAndHoldStrategy(10000,k=2,wait=100,envelope='uniform')
percent_gains = backtest(ts,start="2014-1-1", end="2015-11-02",correct=False)
print('TSE buy and hold  made a %{} return on investment'.format(round(percent_gains,1)) )
```

# Quantifying the Effectiveness of the TSE Buy and Hold Strategy

The TSEBuyAndHoldStrategy seems to be a great performer, lets charecterize how good it is by sampling the space of possible buy and hold strategies, and comparing it to the results in that space. We can do this with the buy_and_hold_spread function defined in the strategies file. This function will randomly sample buy and hold strategies from the space of possible buy and hold strategies (the k ticker subsets). It will then calculate the return for each of these strategies and return them as a list.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from Strategies import buy_and_hold_spread, backtest, TSEBuyAndHoldStrategy
from utils import cache
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor


@cache('data/buy_and_hold_spreads_cache.pkl')
def buy_and_hold_spreads(ks=range(1, 10), wait=100, start="2014-1-1", end="2015-11-02", iterations=300):
    return [buy_and_hold_spread(k, wait, start, end, iterations) for k in ks]

def tse_results(balance=10000,base_model=LinearRegression(), ks=range(1, 10), wait=100, envelope='uniform', start="2014-1-1", end="2015-11-02"):
    strategies = [TSEBuyAndHoldStrategy(balance=balance,base_model=base_model, k=k, wait=wait, envelope=envelope) for k in ks]
    return [backtest(strategy, start=start, end=end, correct=False) for strategy in strategies]

ks = range(1,5)
bh_spreads = buy_and_hold_spreads(ks=ks)
linear_tse_returns = tse_results(ks=ks)
dt_tse_returns = tse_results(base_model=DecisionTreeRegressor(),ks=ks)

linear_probs = [sum(1 if tser > bhr else 0 for bhr in bhs) / float(len(bhs))
         for (tser, bhs) in zip(linear_tse_returns, bh_spreads)]

dt_probs = [sum(1 if tser > bhr else 0 for bhr in bhs) / float(len(bhs))
         for (tser, bhs) in zip(dt_tse_returns, bh_spreads)]


plt.subplot(1, 2, 1)
plt.plot(ks, linear_probs, 'r-',label='linear')
plt.plot(ks, dt_probs, 'g-',label='Decision Tree')
plt.xlabel('k')
plt.ylabel('Percentage of Buy And Hold Strategies Beaten')

plt.subplot(1, 2, 2)
plt.boxplot(bh_spreads)
for i  in range(len(ks)):
    x = np.random.normal(i + 1, 0.02, size=len(bh_spreads[i]))
    if i == 0:
        plt.plot(x, bh_spreads[i], 'bo', alpha=.3, label='BH Gains')
        plt.plot(i + 1, linear_tse_returns[i], 'ro', ms=10, label='Linear TSE Gains')
        plt.plot(i + 1, dt_tse_returns[i], 'go', ms=10, label='Decision Tree TSE Gains')
    else:
        plt.plot(x, bh_spreads[i], 'bo', alpha=.3)
        plt.plot(i + 1, linear_tse_returns[i], 'ro', ms=10)
        plt.plot(i + 1, dt_tse_returns[i], 'go', ms=10)
plt.xlabel('k')
plt.ylabel('Percentage Return')
plt.legend()
plt.gcf().set_size_inches(15,6)
plt.show()
```

As one can see, the TSE Buy and Hold Strategy performs substantially better than the average buy and hold strategy. And seems to perform the best when it only chooses the top performing stock.

# Investigating the Linear TSE's Ability to Choose an Above Average Stock

If our trading strategy is going to be predicated on choosing an above average stock to invest in, lets investigate that the TSE will choose an above average stock most of the time. To do this we will break up a large dataset from 2009-2015 into 20 different smaller datasets using cascade_cv. We will then see if the TSE chooses an above average stock, aka a stock that has a change larger than the mean change. We will also simultaneously investigate how this ability depends on training dataset size.

```python
%matplotlib inline
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from TimeSeriesEstimator import TimeSeriesRegressor, cascade_cv
from utils import load_s_and_p_data, cache


X = load_s_and_p_data(start="2009-1-1", only_close=True)
names = list(X.columns.values)

@cache('data/data_dependence_cache.pkl')
def get_data_dependece(X, data_sizes, folds=20, test_size=30, n_prev=2, log=True):
    bests = np.empty((len(data_sizes), folds, X.shape[1]))
    for i, data_size in enumerate(data_sizes):
        pairs = cascade_cv(len(X), folds, data_size=data_size, test_size=test_size, number=True)
        for j, pair in enumerate(pairs):
            if log:
                print('data size: {} trial {} '.format(data_size, j))
            X_train, X_test = np.array(X.iloc[pair[0], :]), np.array(X.iloc[pair[1], :])
            tsr = TimeSeriesRegressor(LinearRegression(), n_prev=n_prev)
            tsr.fit(X_train)
            fc = tsr.forecast(X_train, len(X_test))

            def changes(X, start=0, end=-1):
                return np.array([X[end, i] - X[start, i] for i in range(X.shape[1])])

            best_is = changes(fc).argsort()[::-1]
            for k in range(X.shape[1]):
                bests[i, j, k] = changes(X_test)[best_is[k]] - np.mean(changes(X_test))

    return bests


data_sizes = [.1, .2, .3, .4]
bests = get_data_dependece(X, data_sizes)

x = [int(ds*X.shape[0]) for ds in data_sizes]
LR = np.transpose(bests[:, :, 0])
SLR = np.transpose(bests[:, :, 1])
plt.plot(x, np.mean(LR, axis=0), color='b', label='Largest Rise')
plt.plot(x, np.mean(SLR, axis=0), color='g', label='Second Largest Rise')

plt.xlabel('Size of dataset')
plt.ylabel('mean(Price Change - mean(Price Change))')
plt.legend()
plt.show()
```

As one can see, a linear time series estimator chooses an above average stock (>0) on average!
