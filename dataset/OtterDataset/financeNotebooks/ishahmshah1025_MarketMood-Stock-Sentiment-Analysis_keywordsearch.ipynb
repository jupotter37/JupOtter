```python
import requests
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
import spacy
```

```python
nltk.download('vader_lexicon')
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def fetch_news_by_keyword():
    api_key = "9eca8bc96b9e453496b85e2ced9b8dfe"  
    keyword = input("Enter a keyword to search for news: ")  
    
    url = f"https://newsapi.org/v2/everything?q={keyword}&apiKey={api_key}"
    
    response = requests.get(url)
    data = response.json()

    if data['status'] == 'ok':
        articles = data['articles']
        if articles:
            sentiment_analyzer = SentimentIntensityAnalyzer()
            total_sentiment = {'bearish': 0, 'neutral': 0, 'bullish': 0}
            num_articles = 0

            for article in articles:
                print(f"Title: {article['title']}")
                print(f"Source: {article['source']['name']}")
                print(f"Published At: {article['publishedAt']}")
                print(f"Description: {article['description']}")
                print(f"URL: {article['url']}\n")

                if article['description']:
                    sentiment_scores = sentiment_analyzer.polarity_scores(article['description'])
                    total_sentiment['bearish'] += sentiment_scores['neg']
                    total_sentiment['neutral'] += sentiment_scores['neu']
                    total_sentiment['bullish'] += sentiment_scores['pos']
                    num_articles += 1

    
            if num_articles > 0:
                avg_bearish = total_sentiment['bearish'] / num_articles
                avg_neutral = total_sentiment['neutral'] / num_articles
                avg_bullish = total_sentiment['bullish'] / num_articles

                print(f"\nAverage Sentiment Scores for '{keyword}':")
                print(f"Bearish: {avg_bearish:.4f}")
                print(f"Neutral: {avg_neutral:.4f}")
                print(f"Bullish: {avg_bullish:.4f}")
        else:
            print("No articles found.")
    else:
        print("Error fetching news articles.")
```

```python
fetch_news_by_keyword()
```

```python
def filter_articles_by_recency(articles):
    sorted_articles = sorted(articles, key=lambda x: x['publishedAt'], reverse=True)

    for article in sorted_articles:
        title = article['title']
        source = article['source']['name']
        published_at = article['publishedAt']
        description = article['description']
        article_url = article['url']

        print(f"Title: {title}")
        print(f"Source: {source}")
        print(f"Published At: {published_at}")
        print(f"Description: {description}")
        print(f"URL: {article_url}\n")
```

```python
def recency_sorted_articles():
    api_key = "9eca8bc96b9e453496b85e2ced9b8dfe"  
    keyword = input("Enter a keyword to search for news to filter: ")  
    
    url = f"https://newsapi.org/v2/everything?q={keyword}&apiKey={api_key}"
    
    response = requests.get(url)
    data = response.json()

    if data['status'] == 'ok':
        articles = data['articles']
        if articles:
            filter_articles_by_recency(articles)
        else:
            print("No articles found.")
    else:
        print("Error fetching news articles.")
```
