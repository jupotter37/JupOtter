## Sentiment Analysis

```python
sid = SentimentIntensityAnalyzer()
```

```python
description1 = articles_df.iloc[0]['description']
description1
```

```python
ss = sid.polarity_scores(description1)
print(ss)
```

### Categories

| Polarity | Category |
| --------- | --------- |
| < -0.5 | Very negative |
| -0.5 > x > 0| Negative |
|0        |  Neutral|
|0 < x < 0.5 | Positive|
| > 0.5 | Very positive|

```python
def get_sentiment(description):    
    polarity = sid.polarity_scores(description)['compound']        
    if(polarity>0.5):
        sentiment_category = 'Very Positive'
    elif(polarity>0):
        sentiment_category = 'Positive'
    elif(polarity==0):
        sentiment_category = 'Neutral'
    elif(polarity>-0.5):
        sentiment_category = 'Negative'
    else:
        sentiment_category = 'Very Negative'
    return sentiment_category
```

```python
articles_df['sentiment'] = articles_df['content'].apply(get_sentiment)
articles_df.head()
```
