```python
import pandas as pd
from pathlib import Path
from api_key import weather_api_key
import time
import requests
```

```python
sightings_file = Path("Output/usa_sightings_df_.csv")
```

```python
sightings_df = pd.read_csv(sightings_file)
sightings_df
```

```python
# dropping duplicates
sightings_df_copy = sightings_df.drop_duplicates(subset=['CITY'])
sightings_df_copy
```

```python
cities = sightings_df_copy['CITY'].to_list()
len(cities)
```

```python
base_url='https://api.openweathermap.org/data/2.5/weather?'
# url = f"{base_url}q={city}&units=imperial&appid={weather_api_key}"
# print(url)
# response = requests.get(url)
# data = response.json()
# data
```

```python
city_data = []

for i, city in enumerate(cities):
    time.sleep(2)
    city_url = f"{base_url}q={city}&units=imperial&appid={weather_api_key}"
    print(city_url)
    req = requests.get(city_url)
    req=req.json()

 # Run an API request for each of the cities
    try:
        #Parse the JSON and retrieve data
        city_weather = req['name']

        #Parse out latitude, longitude, max temp, humidity, cloudiness, wind speed, country, and date
        
        city_lat = req['coord']['lat']
        city_lng = req['coord']['lon']

        #Append the City information into city_data list
        city_data.append({"City": city_weather, 
                          "Lat": city_lat, 
                          "Lng": city_lng, 
                        })

    #If an error is experienced, skip the city
    except Exception as e:
        print(e)
        print("City not found. Skipping...")
        pass
              
# Indicate that Data Loading is complete 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
city_data
```

```python
city_data_df1 = pd.DataFrame(city_data)
city_data_df1
```

```python
city_data_df1.to_csv("city Data.csv")
```

```python
new_usa_sightings_df = pd.merge(city_data_df1, sightings_df, how = "left",  on = ["City", "CITY"] )
new_usa_sightings_df
```
