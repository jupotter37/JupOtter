# **Fairness in Music Recommender Systems: A Mini Project Report**

## **Introduction**

In the age of digital music consumption, recommender systems play a pivotal role in shaping users' listening experiences. These systems influence not only individual preferences but also the visibility and success of artists across the globe. However, biases in recommender systems can lead to unequal representation of artists based on attributes like country, gender, or genre, thereby impacting fairness and diversity in music recommendations.

This mini-project explores fairness in music recommender systems by analyzing potential biases in artist recommendations, particularly focusing on country-based disparities between artists from the United States (US) and Great Britain (GB). The study also examines how bias mitigation strategies can promote fairness, aligning with Sustainable Development Goals (SDGs) related to industry innovation and infrastructure (SDG 9) and peace, justice, and strong institutions (SDG 16).



## **Relation to Sustainability Goals**

### **SDG 9: Industry, Innovation, and Infrastructure**

This project contributes to SDG 9 by improving the reliability and fairness of technological processes in recommender systems. By controlling variables that affect reproducibility and addressing biases, the project enhances the technological infrastructure of music recommendation platforms, fostering innovation and equitable access to cultural content.

## **Objective**

- **Primary Goal:** Investigate potential biases in music recommender systems concerning artist country representation, specifically between US and GB artists.
- **Secondary Goal:** Implement bias mitigation strategies to enhance fairness in recommendations and evaluate their effectiveness.



## **Methodology**

### **Data Collection and Preprocessing**

- **Dataset:** The Last.fm 1K dataset, containing user interactions with music tracks.
- **Sampling:** A 1% random sample of the interactions data to manage computational resources.
- **Data Cleaning:** Removal of invalid or missing `artistID`s and conversion of identifiers to string types.
- **Top-N Artists Selection:** Computed play counts to select the top 1,000 artists for focused analysis.
- **Metadata Enrichment:** Fetched artist metadata (gender and country) from the MusicBrainz database to enrich the interactions data.

### **Fairness Analysis Framework**

1. **Demographic Analysis:**
   - Examined the distribution of artist countries in the interactions data to establish a baseline.
   - Identified the US and GB as the most represented countries.

2. **Definition of Protected Groups:**
   - **Privileged Group:** US artists.
   - **Unprivileged Group:** GB artists.

3. **Recommendation Model:**
   - Employed the Singular Value Decomposition (SVD) algorithm using the Surprise library to generate top-10 recommendations for each user.
   - Trained the model on user-artist play counts.

4. **Bias Measurement:**
   - Calculated the **Statistical Parity Difference (SPD)** to quantify bias in recommendations.
     - **SPD Formula:** \( \text{SPD} = P(\text{recommendation} | \text{unprivileged group}) - P(\text{recommendation} | \text{privileged group}) \)
   - Analyzed the distribution of recommended artists by country.

5. **Bias Mitigation:**
   - Implemented a re-ranking strategy to adjust the proportion of GB artists in recommendations.
   - Aimed for equal representation (50% GB artists) in the top-10 recommendations for each user.
   - Recalculated SPD after re-ranking to assess improvement.

**After unzipping *"lastfm-dataset-1K.tar.gz"*, we convert large TSV file (*userid-timestamp-artid-artname-traid-traname*) to Parquet, in order to enhance storage efficiency, accelerate data processing, and ensure data integrity through schema enforcement.**

````python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# File paths
file_path = r'G:\GitHub\datasci-funtime\data\lastfm-dataset-1K\userid-timestamp-artid-artname-traid-traname.tsv'
parquet_file = r'G:\GitHub\datasci-funtime\data\lastfm-dataset-1K\data.parquet'

# Parameters
chunksize = 10 ** 6
column_names = ['userid', 'timestamp', 'artid', 'artname', 'traid', 'traname']
dtype_dict = {
    'userid': 'str',
    'timestamp': 'str',
    'artid': 'str',
    'artname': 'category',
    'traid': 'str',
    'traname': 'category'
}

# Define a fixed PyArrow schema with int32 for dictionary indices
schema = pa.schema([
    ('userid', pa.string()),
    ('timestamp', pa.timestamp('ns', tz='UTC')),  # Ensure consistent timezone
    ('artid', pa.string()),
    ('artname', pa.dictionary(pa.int32(), pa.string())),  # Changed to int32
    ('traid', pa.string()),
    ('traname', pa.dictionary(pa.int32(), pa.string()))   # Changed to int32
])

# Initialize ParquetWriter with the fixed schema
writer = pq.ParquetWriter(parquet_file, schema, compression='snappy')  # Choose compression as needed

# Initialize a counter for skipped lines (optional)
skipped_lines = 0

try:
    for chunk_number, chunk in enumerate(pd.read_csv(
        file_path,
        sep='\t',
        chunksize=chunksize,
        names=column_names,
        dtype=dtype_dict,
        on_bad_lines='skip',  # Skip malformed lines
        encoding='utf-8',
        header=0  # Assuming the first line is the header
    ), 1):
        # Convert 'timestamp' to datetime with UTC timezone
        chunk['timestamp'] = pd.to_datetime(chunk['timestamp'], errors='coerce', utc=True)

        # Convert the Pandas DataFrame chunk to a PyArrow Table without preserving the index
        table = pa.Table.from_pandas(chunk, preserve_index=False, schema=schema)

        # Write the table chunk to the Parquet file
        writer.write_table(table)

        print(f"Converted and wrote chunk {chunk_number} with {len(chunk)} rows to Parquet")

except pd.errors.ParserWarning as pw:
    skipped_lines += 1
    print(f"ParserWarning: {pw}")
except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Ensure the ParquetWriter is properly closed
    writer.close()
    print("ParquetWriter closed successfully.")

print(f"Conversion to Parquet complete. Total skipped lines: {skipped_lines}")
````

````python
import pandas as pd
import os
import musicbrainzngs
import time
import logging
from surprise import Dataset, Reader, SVD
from collections import defaultdict
import heapq
import matplotlib.pyplot as plt

# -------------------------------------------
# Setup Logging Configuration
# -------------------------------------------
logging.basicConfig(
    filename='recommender_system_debug.log',  # Log output to file
    level=logging.DEBUG,                       # Set logging level (DEBUG, INFO, WARNING, ERROR, CRITICAL)
    format='%(asctime)s - %(levelname)s - %(message)s',  # Log format
    filemode='w'                               # Overwrite the log file each run
)

# -------------------------------------------
# Initialize MusicBrainz User Agent
# -------------------------------------------
musicbrainzngs.set_useragent("MusicRecommenderBiasProject", "1.0", "ash.mathcom@gmail.com")
logging.info("Initialized MusicBrainz user agent.")

# -------------------------------------------
# Define Data Paths
# -------------------------------------------
parquet_file = r'C:\GitHub\datasci-funtime\data\lastfm-dataset-1K\data.parquet'
user_profile_file = r'C:\Github\datasci-funtime\data\lastfm-dataset-1K\userid-profile.tsv'
metadata_cache_file = 'artist_metadata_top_n.csv'
logging.info("Defined data file paths.")

# -------------------------------------------
# Load and Preprocess Interactions Data
# -------------------------------------------
def load_and_preprocess_interactions(parquet_path, sample_frac=0.01, random_state=42):
    logging.info("Loading user interactions data from Parquet file.")
    try:
        interactions_df = pd.read_parquet(parquet_path, engine='pyarrow')
        logging.info(f"Successfully loaded interactions data with {len(interactions_df)} rows.")
    except Exception as e:
        logging.error(f"Failed to load interactions data from Parquet: {e}")
        raise
    
    # Rename columns to match the expected format
    interactions_df.rename(columns={
        'userid': 'userID',
        'artid': 'artistID',
        'artname': 'artistName',
        'traid': 'trackID',
        'traname': 'trackName'
    }, inplace=True)
    logging.debug(f"Columns after renaming: {interactions_df.columns.tolist()}")
    
    # Sample a fraction of the data
    sampled_df = interactions_df.sample(frac=sample_frac, random_state=random_state)
    logging.info(f"Sampled {len(sampled_df)} rows from interactions data (Fraction: {sample_frac}).")
    
    # Convert 'artistID' to string type
    sampled_df['artistID'] = sampled_df['artistID'].astype(str)
    logging.debug("Converted 'artistID' to string type.")
    
    return sampled_df

# Load and preprocess interactions data
sampled_interactions_df = load_and_preprocess_interactions(parquet_file)

# -------------------------------------------
# Compute Play Counts and Select Top-N Artists
# -------------------------------------------
def compute_top_n_artists(interactions_df, top_n=1000, invalid_ids=['None', 'nan']):
    initial_count = len(interactions_df)
    # Remove invalid 'artistID's
    filtered_df = interactions_df[~interactions_df['artistID'].isin(invalid_ids)]
    removed_count = initial_count - len(filtered_df)
    logging.info(f"Removed {removed_count} invalid 'artistID's (None/nan).")
    
    # Group by 'artistID' to get play counts
    artist_play_counts = filtered_df.groupby('artistID').size().reset_index(name='playCount')
    logging.debug(f"Computed play counts for {len(artist_play_counts)} artists.")
    
    # Resolve duplicate 'artistID's by keeping the most frequent 'artistName'
    unique_artists = (
        filtered_df.groupby('artistID')['artistName']
        .agg(lambda x: x.value_counts().idxmax())
        .reset_index()
    )
    logging.info("Resolved duplicate 'artistID's with the most frequent 'artistName'.")
    
    # Merge 'artistName' back into 'artist_play_counts'
    artist_play_counts = artist_play_counts.merge(unique_artists, on='artistID', how='left')
    logging.debug("Merged 'artistName' back into 'artist_play_counts'.")
    
    # Sort artists by play counts in descending order and select top N
    top_artists = artist_play_counts.sort_values(by='playCount', ascending=False).head(top_n).reset_index(drop=True)
    logging.info(f"Selected top {top_n} artists based on play counts.")
    
    # Ensure 'artistID's are valid
    top_artists = top_artists[~top_artists['artistID'].isin(invalid_ids)]
    logging.debug(f"Top artists after removing invalid 'artistID's: {top_artists.head()}")
    
    return top_artists

# Compute top-N artists
TOP_N = 1000
top_artists = compute_top_n_artists(sampled_interactions_df, top_n=TOP_N)

# -------------------------------------------
# Filter Interactions to Top-N Artists
# -------------------------------------------
def filter_interactions_to_top_n(interactions_df, top_artists_df):
    top_artist_ids = set(top_artists_df['artistID'])
    interactions_top_n = interactions_df[interactions_df['artistID'].isin(top_artist_ids)].reset_index(drop=True)
    logging.info(f"Filtered interactions to {len(interactions_top_n)} rows that include only top-{TOP_N} artists.")
    return interactions_top_n

# Filter interactions to top-N artists
interactions_df_top_n = filter_interactions_to_top_n(sampled_interactions_df, top_artists)

# -------------------------------------------
# Fetch Artist Metadata from MusicBrainz
# -------------------------------------------
def fetch_artist_metadata(top_artists_df, cache_file, sleep_time=0.1):
    if os.path.exists(cache_file):
        logging.info("Loading cached artist metadata for top-N artists.")
        try:
            artists_df = pd.read_csv(cache_file)
            logging.info(f"Loaded cached artist metadata with {len(artists_df)} records.")
        except Exception as e:
            logging.error(f"Failed to load cached metadata: {e}")
            raise
    else:
        logging.info("Fetching artist metadata from MusicBrainz for top-N artists.")
        artist_metadata = []
        
        for index, row in top_artists_df.iterrows():
            artist_id = row['artistID']
            artist_name = row['artistName']
    
            try:
                result = musicbrainzngs.search_artists(artist=artist_name, limit=1)
                if result['artist-list']:
                    artist = result['artist-list'][0]
                    gender = artist.get('gender', 'Unknown')
                    country = artist.get('country', 'Unknown')
                    logging.debug(f"Fetched metadata for artist: {artist_name} (ID: {artist_id})")
                else:
                    gender = 'Unknown'
                    country = 'Unknown'
                    logging.warning(f"No metadata found for artist: {artist_name} (ID: {artist_id})")
            except Exception as e:
                logging.error(f"Error fetching data for {artist_name} (ID: {artist_id}): {e}")
                gender = 'Unknown'
                country = 'Unknown'
                time.sleep(1)  # Wait before retrying
            
            artist_metadata.append({
                'artistID': artist_id,
                'artistName': artist_name,
                'gender': gender,
                'country': country
            })
    
            # Log progress every 10 artists
            if (index + 1) % 10 == 0 or (index + 1) == len(top_artists_df):
                logging.info(f'Processed {index + 1}/{len(top_artists_df)} artists.')
    
            # Respect API rate limits
            time.sleep(sleep_time)
        
        # Create a DataFrame from the metadata
        artists_df = pd.DataFrame(artist_metadata)
        logging.info(f"Fetched metadata for {len(artists_df)} artists.")
    
        # Save to CSV for future use
        try:
            artists_df.to_csv(cache_file, index=False)
            logging.info(f"Artist metadata saved to {cache_file}.")
        except Exception as e:
            logging.error(f"Failed to save artist metadata to {cache_file}: {e}")
            raise
    
    return artists_df

# Fetch artist metadata
artists_df = fetch_artist_metadata(top_artists, metadata_cache_file)

# -------------------------------------------
# Merge Artist Metadata with Interactions Data
# -------------------------------------------
def merge_metadata_with_interactions(interactions_df, artists_df):
    merged_df = interactions_df.merge(
        artists_df[['artistID', 'gender', 'country']],
        on='artistID',
        how='left'
    )
    logging.info("Merged artist metadata with interactions data.")
    logging.debug(f"Merged DataFrame head:\n{merged_df.head()}")
    return merged_df

# Merge artist metadata with interactions data
interactions_df_top_n = merge_metadata_with_interactions(interactions_df_top_n, artists_df)

# -------------------------------------------
# Analyze Demographics in Interactions Data
# -------------------------------------------
def analyze_demographics(interactions_df):
    # Calculate proportions for 'gender'
    gender_counts = interactions_df['gender'].value_counts(normalize=True)
    logging.info("Gender Distribution in Interactions:")
    logging.info(f"\n{gender_counts}")
    
    # Calculate proportions for 'country'
    country_counts = interactions_df['country'].value_counts(normalize=True)
    logging.info("Country Distribution in Interactions:")
    logging.info(f"\n{country_counts}")
    
    return gender_counts, country_counts

# Analyze demographics
gender_counts, country_counts = analyze_demographics(interactions_df_top_n)

# -------------------------------------------
# Define Groups for Bias Analysis
# -------------------------------------------
def define_bias_groups():
    privileged_group = 'US'
    unprivileged_group = 'GB'
    logging.info(f"Defined privileged group: {privileged_group}, unprivileged group: {unprivileged_group}.")
    return privileged_group, unprivileged_group

# Define privileged and unprivileged groups
privileged_group, unprivileged_group = define_bias_groups()

# -------------------------------------------
# Generate Recommendations with SVD Model
# -------------------------------------------
def prepare_data_for_modeling(interactions_df):
    # Ensure 'userID' and 'artistID' are strings
    interactions_df['userID'] = interactions_df['userID'].astype(str)
    interactions_df['artistID'] = interactions_df['artistID'].astype(str)
    logging.debug("Ensured 'userID' and 'artistID' are strings.")
    
    # Aggregate play counts per user and artist
    user_artist_plays = interactions_df.groupby(['userID', 'artistID']).size().reset_index(name='playCount')
    logging.info(f"Aggregated play counts per user-artist pair: {len(user_artist_plays)} records.")
    
    return user_artist_plays

def build_training_set(user_artist_plays):
    # Prepare data for Surprise
    reader = Reader(rating_scale=(1, user_artist_plays['playCount'].max()))
    data = Dataset.load_from_df(user_artist_plays[['userID', 'artistID', 'playCount']], reader)
    logging.info("Prepared data for Surprise library.")
    return data

def train_svd_model(data):
    # Build the full training set
    trainset = data.build_full_trainset()
    logging.info("Built full training set for Surprise.")
    
    # Initialize the SVD algorithm
    algo_svd = SVD(random_state=42)
    logging.info("Initialized SVD algorithm.")
    
    # Train the algorithm on the training set
    algo_svd.fit(trainset)
    logging.info("Trained SVD model on the training set.")
    
    return algo_svd, trainset

def get_top_n_recommendations(algo, trainset, n=10):
    top_n = defaultdict(list)
    logging.info("Generating top-N recommendations for each user.")
    
    # Iterate over all users in the training set
    for uid in trainset.all_users():
        raw_uid = trainset.to_raw_uid(uid)
        # Get the items the user has already interacted with
        known_items = set([j for (j, _) in trainset.ur[uid]])
        # Predict ratings for items not yet interacted with
        predictions = []
        for iid in trainset.all_items():
            if iid not in known_items:
                raw_iid = trainset.to_raw_iid(iid)
                est = algo.predict(raw_uid, raw_iid).est
                predictions.append((raw_iid, est))
        # Get the top N predictions
        top_n_items = heapq.nlargest(n, predictions, key=lambda x: x[1])
        # Store the recommendations
        top_n[raw_uid] = top_n_items
        # Log progress every 1000 users
        if (uid + 1) % 1000 == 0:
            logging.info(f"Generated recommendations for {uid + 1} users.")
    
    logging.info("Completed generating top-N recommendations for all users.")
    return top_n

def recommendations_to_df(top_n):
    rows = []
    for uid, user_recs in top_n.items():
        for (iid, est) in user_recs:
            rows.append({
                'userID': uid,
                'artistID': iid,
                'estimated_playCount': est
            })
    recs_df = pd.DataFrame(rows)
    logging.info(f"Converted recommendations to DataFrame with {len(recs_df)} rows.")
    return recs_df

# Prepare data for modeling
user_artist_plays = prepare_data_for_modeling(interactions_df_top_n)

# Build training set
data = build_training_set(user_artist_plays)

# Train SVD model
algo_svd, trainset = train_svd_model(data)

# Generate recommendations
top_n_svd = get_top_n_recommendations(algo_svd, trainset, n=10)

# Convert recommendations to DataFrame
recs_svd = recommendations_to_df(top_n_svd)

# -------------------------------------------
# Proceed with Bias Analysis
# -------------------------------------------
def prepare_data_for_bias_analysis(interactions_df, recs_df, artists_metadata_df):
    # Exclude 'Unknown' countries from interactions data
    interactions_filtered = interactions_df[interactions_df['country'] != 'Unknown'].reset_index(drop=True)
    logging.info(f"Excluded 'Unknown' countries from interactions data: {len(interactions_filtered)} rows remain.")
    
    # Merge artist metadata into recommendations
    recs_merged = recs_df.merge(
        artists_metadata_df[['artistID', 'country']],
        on='artistID',
        how='left'
    )
    logging.info("Merged artist metadata into recommendations.")
    
    # Ensure 'country' is not 'Unknown' in recommendations
    recs_filtered = recs_merged[recs_merged['country'] != 'Unknown'].reset_index(drop=True)
    logging.info(f"Excluded 'Unknown' countries from recommendations: {len(recs_filtered)} rows remain.")
    
    # Verify that there are no 'Unknown' countries
    unique_countries = recs_filtered['country'].unique()
    logging.info(f"Countries in recommendations after filtering: {unique_countries}")
    
    return interactions_filtered, recs_filtered

def calculate_statistical_parity(recs_df, protected_attribute, privileged_group, unprivileged_group):
    total_recs = len(recs_df)
    recs_privileged = recs_df[recs_df[protected_attribute] == privileged_group]
    recs_unprivileged = recs_df[recs_df[protected_attribute] == unprivileged_group]
    p_privileged = len(recs_privileged) / total_recs if total_recs > 0 else 0
    p_unprivileged = len(recs_unprivileged) / total_recs if total_recs > 0 else 0
    spd = p_unprivileged - p_privileged
    logging.info(f"Statistical Parity Difference (SPD) calculated: {spd}")
    logging.info(f"Proportion of {privileged_group} artists in recommendations: {p_privileged:.2%}")
    logging.info(f"Proportion of {unprivileged_group} artists in recommendations: {p_unprivileged:.2%}")
    return spd, p_privileged, p_unprivileged

def calculate_recommendation_distribution(recs_df):
    country_recommendations = recs_df['country'].value_counts(normalize=True)
    logging.info("Recommendation Distribution by Country:")
    logging.info(f"\n{country_recommendations}")
    return country_recommendations

# Prepare data for bias analysis
interactions_filtered, recs_filtered = prepare_data_for_bias_analysis(interactions_df_top_n, recs_svd, artists_df)

# Calculate Statistical Parity Difference (SPD)
spd, p_privileged, p_unprivileged = calculate_statistical_parity(
    recs_df=recs_filtered,
    protected_attribute='country',
    privileged_group=privileged_group,
    unprivileged_group=unprivileged_group
)

def calculate_and_plot_distributions(interactions_df, recs_df, privileged_group, unprivileged_group):
    # Plot the distribution of recommendations by country
    plt.figure(figsize=(12, 7))
    recs_df['country'].value_counts(normalize=True).plot(kind='bar', color='skyblue')
    plt.title('Distribution of Recommended Artists by Country')
    plt.xlabel('Country')
    plt.ylabel('Proportion of Recommendations')
    plt.tight_layout()
    plt.show()
    logging.info("Plotted distribution of recommended artists by country.")
    
    # Baseline distribution (excluding 'Unknown')
    baseline_country_distribution = interactions_df['country'].value_counts(normalize=True)
    
    # Create a DataFrame for comparison
    comparison_df = pd.DataFrame({
        'Baseline': baseline_country_distribution,
        'Recommendations': recs_df['country'].value_counts(normalize=True)
    }).loc[[privileged_group, unprivileged_group]].fillna(0)
    
    # Plot the comparison
    comparison_df.plot(kind='bar', figsize=(10, 7))
    plt.title('Country Distribution: Baseline vs Recommendations')
    plt.xlabel('Country')
    plt.ylabel('Proportion')
    plt.legend(['Baseline', 'Recommendations'])
    plt.tight_layout()
    plt.show()
    logging.info("Plotted comparison of baseline and recommendation distributions.")
    
    return baseline_country_distribution, recs_df['country'].value_counts(normalize=True)

# Plot distributions and compare with baseline
baseline_country_distribution, country_recommendations = calculate_and_plot_distributions(
    interactions_filtered,
    recs_filtered,
    privileged_group,
    unprivileged_group
)

def display_results(spd, p_privileged, p_unprivileged, country_recs):
    print("=== Fairness Analysis Results ===")
    print(f"Statistical Parity Difference (GB vs. US): {spd:.4f}")
    print(f"Proportion of US artists in recommendations: {p_privileged:.2%}")
    print(f"Proportion of GB artists in recommendations: {p_unprivileged:.2%}")
    print("\nRecommendation Distribution by Country:")
    print(country_recs)
    logging.info("Displayed fairness analysis results.")
    
# Display the results
display_results(spd, p_privileged, p_unprivileged, country_recommendations)

# -------------------------------------------
# Bias Mitigation Strategies
# -------------------------------------------
def re_rank_by_country(recommendations_df, target_country, desired_proportion):
    re_ranked_recommendations = []
    logging.info(f"Re-ranking recommendations to adjust {target_country} artists to {desired_proportion*100:.1f}% of recommendations.")
    
    # Iterate over each user
    for user_id, group in recommendations_df.groupby('userID'):
        group = group.copy()
        # Determine the number of recommendations to adjust
        n_total = len(group)
        n_target = int(n_total * desired_proportion)
        n_non_target = n_total - n_target

        # Get recommendations for the target and other countries
        target_recs = group[group['country'] == target_country]
        other_recs = group[group['country'] != target_country]

        # Adjust the number of recommendations
        target_recs = target_recs.head(n_target)
        other_recs = other_recs.head(n_non_target)

        # Combine and maintain order
        re_ranked_group = pd.concat([target_recs, other_recs])
        re_ranked_recommendations.append(re_ranked_group)
    
    re_ranked_df = pd.concat(re_ranked_recommendations).reset_index(drop=True)
    logging.info("Completed re-ranking of recommendations.")
    return re_ranked_df

# Re-rank recommendations to aim for equal representation (50% GB artists)
desired_proportion = 0.5
recs_svd_re_ranked = re_rank_by_country(
    recommendations_df=recs_filtered,
    target_country=unprivileged_group,
    desired_proportion=desired_proportion
)

# Recalculate SPD after re-ranking
def calculate_spd(recommendations_df, protected_attribute, privileged_group, unprivileged_group):
    total_recs = len(recommendations_df)
    if total_recs == 0:
        logging.warning("No recommendations to calculate SPD.")
        return 0, 0, 0
    recs_privileged = recommendations_df[recommendations_df[protected_attribute] == privileged_group]
    recs_unprivileged = recommendations_df[recommendations_df[protected_attribute] == unprivileged_group]
    p_privileged = len(recs_privileged) / total_recs
    p_unprivileged = len(recs_unprivileged) / total_recs
    spd = p_unprivileged - p_privileged
    logging.info(f"Statistical Parity Difference (SPD) after re-ranking: {spd}")
    logging.info(f"Proportion of US artists after re-ranking: {p_privileged:.2%}")
    logging.info(f"Proportion of GB artists after re-ranking: {p_unprivileged:.2%}")
    return spd, p_privileged, p_unprivileged

spd_after, p_privileged_after, p_unprivileged_after = calculate_spd(
    recommendations_df=recs_svd_re_ranked,
    protected_attribute='country',
    privileged_group=privileged_group,
    unprivileged_group=unprivileged_group
)

# Log the new SPD
logging.info(f"SPD after re-ranking (GB vs. US): {spd_after}")
logging.info(f"Proportion of US artists after re-ranking: {p_privileged_after:.2%}")
logging.info(f"Proportion of GB artists after re-ranking: {p_unprivileged_after:.2%}")

# Display SPD after re-ranking
print(f"\nSPD after re-ranking (GB vs. US): {spd_after:.4f}")
print(f"Proportion of US artists after re-ranking: {p_privileged_after:.2%}")
print(f"Proportion of GB artists after re-ranking: {p_unprivileged_after:.2%}")

# Optional: Analyze the re-ranked recommendations
def analyze_re_ranked_recommendations(recs_df):
    # Calculate proportions for 'country'
    country_recs = recs_df['country'].value_counts(normalize=True)
    logging.info("Re-ranked Recommendation Distribution by Country:")
    logging.info(f"\n{country_recs}")
    
    # Plot the distribution of re-ranked recommendations by country
    plt.figure(figsize=(12, 7))
    country_recs.plot(kind='bar', color='lightgreen')
    plt.title('Distribution of Re-ranked Recommended Artists by Country')
    plt.xlabel('Country')
    plt.ylabel('Proportion of Recommendations')
    plt.tight_layout()
    plt.show()
    logging.info("Plotted distribution of re-ranked recommended artists by country.")
    
    return country_recs

# Analyze and plot re-ranked recommendations
re_ranked_country_recs = analyze_re_ranked_recommendations(recs_svd_re_ranked)

# -------------------------------------------
# End of Script
# -------------------------------------------
logging.info("Bias analysis and mitigation completed successfully.")
````
