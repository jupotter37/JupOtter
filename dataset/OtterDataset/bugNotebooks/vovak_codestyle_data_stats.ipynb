```python
def read_tokens(token_file):
    with open(token_file, 'r') as f:
        indices = []
        counts = []
        tokens = []
        for line in f:
            if not ',' in line:
                tokens[-1] += line
                continue
            ind, cnt, token = line.split(',', 2)
            indices.append(ind)
            counts.append(cnt)
            tokens.append(token)
        for i in range(len(tokens)):
            tokens[i] = tokens[i][:-1]
        return np.array(indices), np.array(counts), tokens
```

```python
def path2string(node_types):
    v_indices = list(map(int, node_types.split(' ')))
    v_names = list(map(
        lambda ind: "{}:{}".format(
            nodes.loc[ind]['type'], 
            nodes.loc[ind]['direction']
        ),
        v_indices
    ))
    return ', '.join(v_names)
    
def context2string(context):
    tok1, p, tok2 = context
    return '{}\n{}\n{}'.format(tokens.loc[tok1]['value'], 
                               path2string(paths.loc[p]['nodeTypes']), 
                               tokens.loc[tok2]['value'])
```

```python
entities = pd.read_csv('entities.csv', sep=';')
entities['names'] = entities['names'].map(lambda s: str(s).split(','))
entities['emails'] = entities['emails'].map(lambda s: str(s).split(','))
entities['count'] = np.max([entities['authorCount'], entities['emailCount']], axis=0)
entities = entities.drop(['authorCount', 'emailCount'], axis=1)
```

```python
entities.head()
```

```python
saved_entities = {}

def get_entity_id(author, email):
    if (author, email) in saved_entities:
        return saved_entities[(author, email)]
    possible_authors, possible_emails = np.arange(len(entities)), np.arange(len(entities))
    if author not in ['unknown', 'no_reply@jetbrains.com']:
        possible_authors = entities[entities['names'].map(lambda lst: author in lst)].index
    if email not in ['no_reply@jetbrains.com']:
        possible_emails = entities[entities['emails'].map(lambda lst: email in lst)].index
    intersection = np.intersect1d(possible_authors, possible_emails)
    if len(intersection) != 1:
        print(intersection)
        print(author, email)
        print(possible_authors, possible_emails)
    assert len(intersection) == 1
    saved_entities[(author, email)] = intersection[0]
    return intersection[0]
```

```python
counts = np.zeros(len(entities), dtype=np.int)

for file_id in range(data_count):
    print("Processing #{}".format(file_id + 1))
    df = pd.read_csv(data_file.format(file_id))
    for index, row in df.iterrows():
        counts[get_entity_id(row['authorName'], row['authorEmail'])] += 1
    del df
```

```python
buckets = [(1, 2), (3, 4), (5, 6), (7, 10), (11, 20), (21, 30), (31, 40), (41, 50), (51, 100), (101, 500)]
fout = {b: open('changes/changes_{}_{}.csv'.format(b[0], b[1]), 'w') for b in buckets}
```

```python
def get_bucket(k):
    for b in buckets:
        if b[0] <= k <= b[1]:
            return b
```

```python
entity2ids = {entity: [] for entity in changes_by_entity}

for file_id in range(data_count):
    print("Processing #{}".format(file_id + 1))
    df = pd.read_csv(data_file.format(file_id))
    for index, row in df.iterrows():
        if not valid[row['changeId']]:
            continue
        entity = saved_entity(row['authorName'], row['authorEmail'])
        bucket = get_bucket(entity2k[entity])
        cnt += 1
        entity2ids[entity].append(cnt)
        fout[bucket].write('{},{},{},{},{},{},{},{},{}\n'.format(cnt, 
                                                   row['changeId'], 
                                                   entity, 
                                                   row['methodBeforeId'], row['methodAfterId'], 
                                                   row['pathsCountBefore'], row['pathsCountAfter'], 
                                                   row['pathsBefore'], row['pathsAfter']))
    del df
```

```python
cnt = 0

entity2ids = {entity: [] for entity in changes_by_entity}

for file_id in range(data_count):
    print("Processing #{}".format(file_id + 1))
    df = pd.read_csv(data_file.format(file_id))
    for index, row in df.iterrows():
        if not valid[row['changeId']]:
            continue
        entity = saved_entity(row['authorName'], row['authorEmail'])
        bucket = get_bucket(entity2k[entity])
        cnt += 1
        entity2ids[entity].append(cnt)
    del df
```

```python
fout_train = {b: open('batches/train_batches_{}_{}.csv'.format(b[0], b[1]), 'w') for b in buckets}
fout_test = {b: open('batches/test_batches_{}_{}.csv'.format(b[0], b[1]), 'w') for b in buckets}
```

```python
np.random.seed(42)
test_size = 0.3
train_size = 1 - test_size
batch_multiplier = 1
batch_size = 16

for i, entity in enumerate(entity2ids):
    print('Processing {}/{}'.format(i + 1, len(entity2ids)))
    k = entity2k[entity]
    bucket = get_bucket(k)
    ids = entity2ids[entity]
    np.random.shuffle(ids)
    train_ids = ids[:int(len(ids) * train_size)]
    test_ids = ids[len(train_ids):]
    for _ in range(len(train_ids) * batch_multiplier):
        indices = np.random.choice(train_ids, batch_size)
        fout_train[bucket].write(str(entity) + ',' + ','.join(str(ind) for ind in indices) + '\n')
            
    for _ in range(len(test_ids) * batch_multiplier):
        indices = np.random.choice(test_ids, batch_size)
        fout_test[bucket].write(str(entity) + ',' + ','.join(str(ind) for ind in indices)+ '\n')
```
