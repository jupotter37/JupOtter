```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QAction, qApp, QWidget
from PyQt5.QtGui import QIcon

class MyApp(QMainWindow):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        exitAction = QAction(QIcon('exit.png'), 'Exit', self)
        exitAction.setShortcut('Ctrl+Q')
        exitAction.setStatusTip('Exit application')
        exitAction.triggered.connect(qApp.quit)

        self.statusBar()

        menubar = self.menuBar()
        menubar.setNativeMenuBar(False)
        filemenu = menubar.addMenu('&File')
        filemenu.addAction(exitAction)

        self.setWindowTitle('Menubar')
        self.setGeometry(300, 300, 300, 200)
        self.show()

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```

```python
# QT Designer 로 생성한 윈도우 창을 여는 방법

import sys
from PyQt5.QtWidgets import *
from PyQt5 import uic

form_class = uic.loadUiType("mywindow4.ui")[0]

class MyWindow(QMainWindow, form_class):

    def __init__(self):
        super().__init__()
        self.setupUi(self)

        self.pushButton_1.clicked.connect(self.btn1_clicked)
        self.pushButton_2.clicked.connect(self.btn2_clicked)

        self.radioButton_1.clicked.connect(self.radioFunction)
        self.radioButton_2.clicked.connect(self.radioFunction)
        self.radioButton_3.clicked.connect(self.radioFunction)

        self.check1.clicked.connect(self.checkBoxFunction)
        self.check2.clicked.connect(self.checkBoxFunction)
        self.check3.clicked.connect(self.checkBoxFunction)
        self.checkBox1.clicked.connect(self.checkBoxFunction)
        self.checkBox2.clicked.connect(self.checkBoxFunction)
        self.checkBox3.clicked.connect(self.checkBoxFunction)

    def btn1_clicked(self):
        print("클릭 1 확인")

    def btn2_clicked(self):
        print("클릭 2 확인")

    def radioFunction(self):
        if self.radioButton_1.isChecked():
            print("radio button 1 Checked")
        if self.radioButton_2.isChecked():
            print("radio button 2 Checked")
        if self.radioButton_3.isChecked():
            print("radio button 3 Checked")

    def checkBoxFunction(self):
        if self.check1.isChecked():
            print("check1 Checked")
        if self.check2.isChecked():
            print("check2 Checked")
        if self.check3.isChecked():
            print("check3 Checked")
        if self.checkBox1.isChecked():
            print("checkBox1 Checked")
        if self.checkBox2.isChecked():
            print("checkBox2 Checked")
        if self.checkBox3.isChecked():
            print("checkBox3 Checked")

app = QApplication(sys.argv)
window = MyWindow()
window.show()
app.exec_()
```

```python
# 라벨 처리

import sys
from PyQt5.QtWidgets import *
from PyQt5 import uic

form_class = uic.loadUiType("mywindow8.ui")[0]

class MyWindow(QMainWindow, form_class):

    def __init__(self):
        super().__init__()
        self.setupUi(self)

        self.pushButton_1.clicked.connect(self.btn1_clicked)
        self.pushButton_2.clicked.connect(self.btn2_clicked)

        self.radioButton_1.clicked.connect(self.radioFunction)
        self.radioButton_2.clicked.connect(self.radioFunction)
        self.radioButton_3.clicked.connect(self.radioFunction)

        self.check1.stateChanged.connect(self.chkFunction)
        self.check2.stateChanged.connect(self.chkFunction)
        self.check3.stateChanged.connect(self.chkFunction)

        self.checkBox1.stateChanged.connect(self.chkBoxFunction)
        self.checkBox2.stateChanged.connect(self.chkBoxFunction)
        self.checkBox3.stateChanged.connect(self.chkBoxFunction)

        self.pushButton_3.clicked.connect(self.melonFucntion)
        self.pushButton_4.clicked.connect(self.genieFucntion)

        self.pushButton_5.clicked.connect(self.deleteFucntion)
        self.pushButton_6.clicked.connect(self.printFucntion)

        self.printBtn.clicked.connect(self.printBtnFucntion)
        self.setTextBtn.clicked.connect(self.setTextBtnFucntion)
        self.appendTextBtn.clicked.connect(self.appendTextBtnFucntion)
        self.clearBtn.clicked.connect(self.clearBtnFucntion)

        self.lineEdit.textChanged.connect(self.lineEditFuntion)
        self.lineEdit.returnPressed.connect(self.returnFuntion)
        self.changeBtn.clicked.connect(self.changeFuntion)

    def btn1_clicked(self):
        print("클릭 1 확인")

    def btn2_clicked(self):
        print("클릭 2 확인")

    def radioFunction(self):
        if self.radioButton_1.isChecked():
            print("radio button 1 Checked")
        if self.radioButton_2.isChecked():
            print("radio button 2 Checked")
        if self.radioButton_3.isChecked():
            print("radio button 3 Checked")

    def chkFunction(self):
        if self.check1.isChecked():
            print("check1 Checked")
        if self.check2.isChecked():
            print("check2 Checked")
        if self.check3.isChecked():
            print("check3 Checked")

    def chkBoxFunction(self):
        if self.checkBox1.isChecked():
            print("checkBox1 Checked")
        if self.checkBox2.isChecked():
            print("checkBox2 Checked")
        if self.checkBox3.isChecked():
            print("checkBox3 Checked")

    def melonFucntion(self):
        self.label.setText("멜론 차트 조회")

    def genieFucntion(self):
        self.label.setText("지니 차트 조회")

    def deleteFucntion(self):
        self.label.clear()

    def printFucntion(self):
        print(self.label.text())
        
    def printBtnFucntion(self):
        print(self.textBrowser.toPlainText())

    def setTextBtnFucntion(self):
        self.textBrowser.setPlainText("setText")

    def appendTextBtnFucntion(self):
        self.textBrowser.append("appendText")

    def clearBtnFucntion(self):
        self.textBrowser.clear()

    def lineEditFuntion(self):
        self.label_text.setText(self.lineEdit.text())

    def returnFuntion(self):
        print(self.lineEdit.text())

    def changeFuntion(self):
        self.lineEdit.setText('글자 변경')

app = QApplication(sys.argv)
window = MyWindow()
window.show()
app.exec_()
```
