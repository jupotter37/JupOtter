```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.parse

def google_search(query, num_results=10):
    results = []
    for start in range(0, num_results, 10):  # Google displays 10 results per page
        url = f"https://www.google.com/search?q={query}&start={start}"
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
        }
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        for g in soup.find_all('h3'):
            parent = g.find_parent('a')
            if parent and 'href' in parent.attrs:
                # Extract the full URL from the href attribute
                href = parent['href']
                parsed_url = urllib.parse.urlparse(href)

                # Extract the actual URL from the query parameters
                params = urllib.parse.parse_qs(parsed_url.query)
                if 'url' in params:
                    link = params['url'][0]  # Get the actual URL
                else:
                    link = href  # Fallback if 'url' parameter is not found
                
                title = g.get_text()  # Get the title text
                results.append((title, link))  # Store as a tuple
                if len(results) >= num_results:  # Limit to specified number of results
                    return results
                
    return results

# Define the websites and their respective search queries for the term "career"
websites = {
    "SearchITManagement": "site:https://techtarget.com career",
    "CIO.com": "site:https://cio.com career",
    "ITProPortal": "site:https://itproportal.com career",
    "InformationWeek": "site:https://informationweek.com career",
    "Harvard Business Review": "site:https://hbr.org career"
}

# Collect results in a list
all_results = []

for name, query in websites.items():
    print(f"Searching for: {name}")
    results = google_search(query, num_results=10)  # You can adjust the number of results if needed
    for title, link in results:
        all_results.append({"Website": name, "Title": title, "Link": link})

# Create a DataFrame for better visualization
results_df = pd.DataFrame(all_results)

# Display the results in a table format
print(results_df)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from openai import OpenAI
from deep_translator import GoogleTranslator
import time
from requests.exceptions import RequestException

# Initialize OpenAI client ADD REAL KEY 
client = OpenAI(api_key='')

def fetch_article_content(url, max_retries=3):
    for attempt in range(max_retries):
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')
            paragraphs = soup.find_all('p')
            article_content = ' '.join([para.get_text() for para in paragraphs])
            return article_content
        except RequestException as e:
            if attempt == max_retries - 1:
                print(f"Failed to fetch content from {url}: {str(e)}")
                return ""
            time.sleep(2 ** attempt)

def decrease_title_number(title):
    try:
        return re.sub(r'(\d+)', lambda x: str(int(x.group(0)) - 1), title)
    except ValueError:
        print(f"Could not process number in title: {title}")
        return title

def rewrite_content(content, max_retries=3):
    for attempt in range(max_retries):
        try:
            response = client.chat.completions.create(
                model="gpt-4o-mini",  # Changed to a standard model name
                messages=[
                    {"role": "system", "content": "You are a helpful assistant that rewrites articles while maintaining their meaning."},
                    {"role": "user", "content": f"Rewrite the following text in a unique way while preserving its meaning: {content}"}
                ],
                temperature=0.7
            )
            return response.choices[0].message.content
        except Exception as e:
            if attempt == max_retries - 1:
                print(f"Failed to rewrite content: {str(e)}")
                return content
            time.sleep(2 ** attempt)

'''def translate_text(text, max_retries=3):
    for attempt in range(max_retries):
        try:
            chunks = [text[i:i+5000] for i in range(0, len(text), 5000)]
            translated_chunks = []
            
            for chunk in chunks:
                translator = GoogleTranslator(source='en', target='ru')
                translated_chunk = translator.translate(chunk)
                translated_chunks.append(translated_chunk)
                time.sleep(1)
            
            return ' '.join(translated_chunks)
        except Exception as e:
            if attempt == max_retries - 1:
                print(f"Failed to translate text: {str(e)}")
                return text
            time.sleep(2 ** attempt)'''

def translate_text(text, max_retries=3):
    """
    Translate text while handling length limitations properly
    """
    try:
        # Check if text is too long
        if len(text) > 5000:
            # Split into smaller chunks of ~4000 characters (leaving room for sentences to complete)
            words = text.split()
            chunks = []
            current_chunk = []
            current_length = 0
            
            for word in words:
                # Add length of word plus space
                if current_length + len(word) + 1 < 4000:
                    current_chunk.append(word)
                    current_length += len(word) + 1
                else:
                    # Save current chunk and start new one
                    chunks.append(' '.join(current_chunk))
                    current_chunk = [word]
                    current_length = len(word) + 1
            
            # Add the last chunk if it exists
            if current_chunk:
                chunks.append(' '.join(current_chunk))
        else:
            chunks = [text]

        translated_chunks = []
        translator = GoogleTranslator(source='en', target='ru')
        
        # Translate each chunk
        for chunk in chunks:
            try:
                translated_chunk = translator.translate(chunk)
                translated_chunks.append(translated_chunk)
                # Print progress for verification
                print(f"Translated chunk of length {len(chunk)}")
                time.sleep(1)  # Rate limiting
            except Exception as e:
                print(f"Error translating chunk: {str(e)}")
                translated_chunks.append(chunk)  # Keep original if translation fails
        
        # Join all translated chunks
        return ' '.join(translated_chunks)
        
    except Exception as e:
        print(f"Translation error: {str(e)}")
        return text

# Load the Excel file
output_file = 'search_results.xlsx'
df = pd.read_excel(output_file)

results = []

# Process each title and link
for index, row in df.iterrows():
    try:
        print(f"Processing article {index + 1}/{len(df)}")
        
        title = row['Title']
        link = row['Link']
        
        article_content = fetch_article_content(link)
        if not article_content:
            continue
        
        new_title = decrease_title_number(title)
        rewritten_content = rewrite_content(article_content)
        translated_content = translate_text(rewritten_content)
        
        results.append({
            'New Title': new_title,
            'Translated Content': translated_content
        })
        
        time.sleep(2)
        
    except Exception as e:
        print(f"Error processing row {index}: {str(e)}")
        continue

# Create and save results
results_df = pd.DataFrame(results)
results_df.to_excel('rewritten_articles.xlsx', index=False)
print("Articles have been processed successfully!")
```
