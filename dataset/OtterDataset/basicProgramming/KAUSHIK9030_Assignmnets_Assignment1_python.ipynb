##5. Explain the concept of type casting in python with examples.


Type casting or type conversion is the process of changing the data type of a value/object in python.

**(a)** **Implicit Type casting** -  It is a built-in function where python automatically converts one data type into another when required, as python understands the data type.

example:

a = 5
b = 3.0

result = a/b    #1.0

type(result)  # float



**(b)** **Explicit Type casting** -  This type of type casting is done manually, using the built-in functions like - (str, int, float, bool).

example:


# Converting float to int

num = 5.67

num_int = int(num)

print(num_int)  # Output:

print(type(num_int))  # Output: <class 'int'>

# Converting int to float

num_float = float(num_int)

print(num_float)  # Output: 5.0

print(type(num_float))  # Output: <class
'float'>

# Converting int to string

num_str = str(num_int)

print(num_str)  # Output: '5'

print(type(num_str))  # Output: <class 'str'>

# Converting list to tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)
print(type(my_tuple))  # Output: <class 'tuple'>
