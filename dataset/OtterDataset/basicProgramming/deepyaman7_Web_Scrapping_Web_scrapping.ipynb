```python
import requests
from bs4 import BeautifulSoup
r = requests.get("https://www.infoworld.com/article/3204016/what-is-python-powerful-intuitive-programming.html")
c = r.content
#print(c)
```

```python
soup = BeautifulSoup(c, 'html.parser')
soup.prettify
```

```python
head = soup.find_all("head")
title = head[0].find_all("title")
title_text = title[0].text
```

```python
title_text
```

```python
body = soup.find_all("body", id = "article3204016")
```

```python
p = body[0].find_all("p")
p_text = p[0:42]
```

```python
for i in p_text:
    #print(i.text)
    file1 = open("MyFile.txt", "a")
    L = i.text
    file1.writelines(L)
    
        
```
