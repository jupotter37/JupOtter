```python
   #Answer 12

"""
The 'elif' statement in python is short for 'else if'  and is used to test multiple condition sequentially
after an initial 'if' condition, it provides a way to add additional conditional checks when the
'if' condition is false.

Steps in 'elif' statement:
1. Evaluate the 'if' condition: if it is true, the code block under the 'if' statement is executed.
2. If the 'if' condition is false, the 'elif' condition is evaluated.
3. If the 'elif' condition is true, the code block under the 'elif' statement is executed.
4. If none of the conditions are true, the code block under the 'else' statement (if present) is executed.

"""
#Example
age = int(input("enter your age: ")) # input age to be checked
if age < 18: # condition 1: If age is below 18 years
  print("you are a minor") # this code will be executed if condition 1 is true
elif age >= 18 and age < 60: # condition 2: if condition 1 is false program will check condition 2
  print("you are an adult") # this code will be executed if condition 2 is true
else:
  print("you are a senior citizen")# this code will be executed if condition 2 is false
```

```python
#Answer 13
"""
The main difference between 'for' and 'while' loop lies in how they operate and the scenarios where each is
typically used.

'for'loop is used when the number of iterations is known or when iterating over a sequence like list, range, etc
in a controlled manner.
"""
# Example
for i in range(5):
  print(i)

"""
'while' loop is used when the number of iterations is not predetermined and depends on a condition being true.
'while' loop continues to execute as long as the condition is true.
"""
# Example
i = 0
while i < 5:
  print(i)
  i += 1
```

```python
# Answer 14
"""
A scenario in which a user has to enter password, we don,t know how many attempts the user will need to input
a valid password,so a 'while' loop is used.
"""
# Example
password = "password123" # correct password
a = str(input("enter password: ")) # input password
while a != password:
  print("incorrect password")
  a = str(input("enter password: "))
print("correct password")
"""
A 'for' loop is used when the number of iterations is known in this case the there is no limit on the number
of attempts the user will need to input a valid password.
"""
```
