```python
#inv-app-example-1.py

"""
==============================
problem stementment
Program requirements
==============================

#1. Write a program to compute four invoice/inventory items.

example:
item1 5
item2 10.5
item3 12
item4 15.5

total ???

"""

#-------------------------------
#--- program steps/features ----
#-------------------------------

#- memory varibles
#- input
#- output
#- calculations
#- decisions - n/a
#- looping - n/a

item1 = 5  #integer, no decimal point
item2 = 10.5 #float, decimal point
item3 = 12  #integer, no decimal point
item4 = 15.5 #float, decimal point

#calculation
total = item1 + item2 + item3 + item4

print("invoice example app 1.0")

print("total: ",total)


```

```python
#inv-app-example-2.py

"""
==============================
problem stementment
Program requirements
==============================

#1. Write a program to compute four invoice/inventory items.

example:

item1 "5"      -- string
item2 "10.5"   -- string
item3 12       -- integer
item4 15.5     -- float

total ???

"""

#-------------------------------
#--- program steps/features ----
#-------------------------------

#- memory varibles
#- input
#- output
#- calculations
#- casting/converting
#- decisions - n/a
#- looping - n/a
#- other

item1 = "5"  #string
item2 = "10.5" #string
item3 = 12  #integer, no decimal point
item4 = 15.5 #float, decimal point

#calculation
#cast data types
total = int(item1) + float(item2) + item3 + item4

print("invoice example app 1.0")

print("total: ",total)

```

```python
#inv-app-example-3.py

"""
==============================
problem stementment
Program requirements
==============================

#1. Write a program to compute four invoice/inventory items.

example:

prompt to input item1
prompt to input item2
prompt to input item3
prompt to input item4

total ???

"""

#-------------------------------
#--- program steps/features ----
#-------------------------------

#- memory varibles
#- input
#- output
#- calculations
#- casting/converting
#- decisions - n/a/
#- looping - n/a
#- other

item1 = input("item1: ")  #string
item2 = input("item2: ")  #string
item3 = input("item3: ")  #string
item4 = input("item4: ")  #string

#calculation
#cast data types
total = float(item1) + float(item2) + float(item3) + float(item4)

print("invoice example app 1.0")

print("total: ",total)

```
