
An API is essentially a set of rules that defines how one software component interacts with another. For example, when you want to get information from a web service, your program (the client) sends a request to the server using an API, and the server responds with the requested data.

2. Request-Response Cycle
Request:

The client sends a request to the server using HTTP. The request typically contains:
URL: The endpoint where the request is sent.
Method: Type of request (e.g., GET, POST).
Headers: Metadata like authentication tokens or content type.
Body: The actual data sent with the request (optional, depending on method).

Response:

The server processes the request and responds with:
Status code: Indicates whether the request was successful (e.g., 200 OK, 404 Not Found).
Headers: Information about the response (e.g., content type).
Body: The data requested, typically in JSON format.



3. Steps to Fetch Data Using Python
Step 1: Identify the API and review the documentation to understand the URL, request methods, parameters, and authentication requirements.
Step 2: Set up the request in Python by specifying the URL, adding headers or authentication keys if needed.
Step 3: Send the request using Python’s HTTP libraries.
Step 4: Receive the response and check the status code to ensure the request was successful.
Step 5: Extract the data from the response body (usually JSON) and parse it for further use.


APIs make it easy to retrieve data from remote services without needing to understand the underlying implementation, making them an essential tool for modern software development.

```python
#Web scraping 

import requests
from bs4 import BeautifulSoup

# Step 1: Define the URL to scrape (example: GSMArena search page for iPhone 16)
url = 'https://www.gsmarena.com/results.php3?sQuickSearch=yes&sName=iphone+16'

# Step 2: Make an HTTP request to the website
response = requests.get(url)

# Step 3: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Step 4: Find relevant information (e.g., news articles, reviews, etc.)
# On GSMArena, articles are usually under a class like 'article-info'
articles = soup.find_all('div', class_='makers')

# Step 5: Loop through the articles and print the titles and links
for article in articles:
    phones = article.find_all('li')
    for phone in phones:
        phone_name = phone.find('span').text
        phone_link = phone.find('a')['href']
        print(f"Phone Name: {phone_name}")
        print(f"Link: https://www.gsmarena.com/{phone_link}")
        print('-' * 50)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Amazon iPhone Search URL
url = 'https://www.amazon.in/s?k=iphone+16'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Send request
response = requests.get(url, headers=headers)
print(response)
soup = BeautifulSoup(response.content, 'html.parser')
#print(soup)

# # Extracting product details
products = soup.find_all('div', {'data-component-type': 's-search-result'})
#print(products)
data = []
for i in products:
    title = i.h2.text.strip()
    price = i.find('span', 'a-price-whole')
    rating = i.find('span', 'a-icon-alt')

    price = price.text.strip() if price else 'N/A'
    rating = rating.text.strip() if rating else 'N/A'

    data.append({
        'Product': title,
        'Price': price,
        'Rating': rating
    })

# Convert to DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = 'http://books.toscrape.com/'


response = requests.get(url)
print(response)
#print(response.content)

if response.status_code == 200:
    
    soup = BeautifulSoup(response.content, 'html.parser')
    books = soup.find_all('article', class_='product_pod')
    titles = []
    prices = []
    availability = []
    
    for i in books:
        title = i.h3.a['title']
        titles.append(title)
        
        
        price = i.find('p', class_='price_color').text
        prices.append(price)
        
        
        avail = i.find('p', class_='instock availability').text.strip()
        availability.append(avail)
    
   
    df = pd.DataFrame({
        'Title': titles,
        'Price': prices,
        'Availability': availability
    })
    
  
    print(df.head(20))
else:
    print('Failed to retrieve data:', response.status_code)
```
