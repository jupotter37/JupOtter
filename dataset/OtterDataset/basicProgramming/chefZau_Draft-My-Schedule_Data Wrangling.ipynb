```python
import os
import json
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
CONFIG = json.load(open('resources/config.json'))
```

```python
DRIVER_PATH = os.path.abspath('./chromedriver')
driver = webdriver.Chrome(DRIVER_PATH)
```

```python
url = 'https://draftmyschedule.uwo.ca/login.cfm'
```

```python
driver.get(url)
```

```python
try: 
    userNameBox = driver.find_element_by_xpath('//input[@name="txtUsername"]')
    passwordBox = driver.find_element_by_xpath('//input[@name="txtPassword"]')

    userNameBox.send_keys(CONFIG['USER'])
    passwordBox.send_keys(CONFIG['PASSWORD'])

    driver.find_element_by_xpath('//button[@type="submit"]').click()
    driver.find_element_by_xpath('//input[@id="agreement"]').click()
    driver.find_element_by_xpath('//button[@type="submit"]').click()
except:
    pass
```

## Extract Data

```python
subjectDropDown = driver.find_element_by_xpath('//select[@id="Subject"]')
```

```python
subject = Select(subjectDropDown)
subjectCount = len(subject.options)
```

```python
results = list()
```

```python
for i in range(1, subjectCount):
    
    try: 
        subjectDropDown = driver.find_element_by_xpath('//select[@id="Subject"]')

        subject = Select(subjectDropDown)
        subject.select_by_index(i)

        subWebElement = subject.first_selected_option
        subjectTitle = subWebElement.text
        subjectValue = subWebElement.get_attribute('value')
        
        print('{')
        print(subjectTitle)

        # the value is the class of the course
        course = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, subjectValue))
        )

        courseCount = len(course)

        for j in range(courseCount):

            course = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, subjectValue))
            )
            course[j].click()
            
            submitButtonPath = '//button[@type="submit"][@class="btn btn-info"]'

            driver.find_element_by_xpath(submitButtonPath).click()

            courseTitle = driver.find_element_by_css_selector('div.clearfix + hr + h4').text
            courseDesc = driver.find_element_by_css_selector('div.clearfix + hr + h4 + p.font-md').text
            
            print(courseTitle)
            
            trPath = '//table[@class="table table-condensed table-hover"]/tbody/tr'
            trs = driver.find_elements_by_xpath(trPath)
            trs = trs[1:]  # skip the headers

            for tr in trs:
                tds = tr.find_elements_by_tag_name('td')
                daysTimesLocation = tr.find_element_by_xpath('//table[@class="table table-bordered table-condensed"]').text
                results.append([subjectTitle, courseTitle, courseDesc] + list(map(lambda x: x.text, tds)) + [daysTimesLocation])
    except:
        continue

    finally:
        print('}')
```

```python
filteredData = list(map(lambda x: x[:8] + x[-8:], results))
```

```python
df = pd.DataFrame(filteredData, columns=[
    'Subject',
    'CourseTitle',
    'CourseDescription',
    'Component',
    'Section',
    'ClassNbr',
    'Instructor',
    'RequisitesAndConstraints',
    'CreditUnits',
    'Status',
    'Waitlist',
    'Campus',
    'DeliveryType',
    'SaveToDraft',
    'Remove',
    'Days/Times/Location',
])
```

```python
weekdayDict = {
    'M': 'Monday',
    'Tu': 'Tuesday',
    'W': 'Wednesday',
    'Th': 'Thursday',
    'F': 'Friday'
}
```

```python
def createJSON(x):
    
    try:
        days = x.split('\n')

        option = list()

        for day in days:

            time = re.search(r'\d+:\d+ (AM|PM) - \d+:\d+ (AM|PM)', day).group(0)
            start, end = time.split(' - ')

            firstDigitIndex = re.search(r"\d", day).start()
            weekdays = day[:firstDigitIndex]
            weekdays = (' '.join(weekdays.split())).split()

            for d in weekdays:

                option.append({
                    "day": weekdayDict[d],
                    "start": start,
                    "end": end
                })

        return option
    except:
        return None
```

```python
df['Times'] = df['Days/Times/Location'].apply(createJSON)
```
