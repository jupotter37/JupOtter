```python
#Q3. Compare and contrast mutable and immutable objects in Python with examples.
```

```python
#In Python, objects are categorized into mutable and immutable types based on whether their state (value) can be modified after they are created. Understanding the difference between mutable and immutable objects is crucial for effective programming, as it affects how data is managed and manipulated in your code.
#Mutable Objects
#Mutable objects are those whose state or value can be changed after they are created. Changes to mutable objects do not create a new object but modify the existing one. This means that if multiple variables reference the same mutable object, changes made through one variable will be reflected in all references.
#Examples of Mutable Objects:
#Lists:
# Creating a mutable list
my_list = [1, 2, 3]
print(my_list)  # Output: [1, 2, 3]

# Modifying the list
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

#2. Dictionaries:
# Creating a mutable dictionary
my_dict = {'a': 1, 'b': 2}
print(my_dict)  # Output: {'a': 1, 'b': 2}

# Modifying the dictionary
my_dict['c'] = 3
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}

#3. Sets:
# Creating a mutable set
my_set = {1, 2, 3}
print(my_set)  # Output: {1, 2, 3}

# Modifying the set
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

#Immutable Objects
#Immutable objects are those whose state or value cannot be changed once they are created. Any operation that attempts to modify an immutable object will instead create a new object. If multiple variables reference the same immutable object, changes made through one variable will not affect others.
#Examples of Immutable Objects:
#Integers:
# Creating an immutable integer
x = 10
print(x)  # Output: 10

# Attempting to modify the integer
x = x + 1
print(x)  # Output: 11
# Note: The original integer 10 remains unchanged; x now references a new integer object.

#2. Strings:
# Creating an immutable string
s = "Hello"
print(s)  # Output: Hello

# Attempting to modify the string
s = s + " World"
print(s)  # Output: Hello World
# Note: The original string "Hello" remains unchanged; s now references a new string object.

#3. Tuples:
# Creating an immutable tuple
t = (1, 2, 3)
print(t)  # Output: (1, 2, 3)

# Attempting to modify the tuple
# t[0] = 4  # This will raise a TypeError

#Comparison:
#Modification:
#Mutable Objects: Can be modified in place (e.g., adding items to a list or dictionary).
#Immutable Objects: Cannot be modified; any change results in the creation of a new object (e.g., modifying a string results in a new string).
#Behavior with References:
#Mutable Objects: Changes to the object through one reference affect all references to that object.
#Immutable Objects: Changes through one reference do not affect other references; each change creates a new object.
#Use Cases:
#Mutable Objects: Suitable for collections of items where modifications are frequent, such as managing a list of users or a dictionary of configurations.
#Immutable Objects: Suitable for fixed data that should not change, such as keys in a dictionary or constant values.
#Performance Considerations:
#Mutable Objects: Can be more efficient for scenarios where frequent modifications are needed, as they avoid the overhead of creating new objects.
#Immutable Objects: Can be more predictable and safer in concurrent programming scenarios since their state cannot change unexpectedly.
#Understanding these differences helps in choosing the appropriate type for data structures and algorithms, ensuring that your code behaves as expected and performs efficiently
```

```python
#Q4. Discuss the different types of operators in Python and provide examples of how they are used.
```

```python
#In Python, operators are special symbols used to perform operations on variables and values. They can be categorized into several types based on their functionality. Here’s a detailed overview of each type of operator and examples of how they are used:
#1. Arithmetic Operators
#Arithmetic operators are used for basic mathematical operations.
#Addition (+): Adds two values.
#code
a = 10
b = 5
result = a + b  # result is 15


#Subtraction (-): Subtracts the second value from the first value.
#code
result = a - b  # result is 5


#Multiplication (*): Multiplies two values.
#code
result = a * b  # result is 50


#Division (/): Divides the first value by the second value, returns a float.
#code
result = a / b  # result is 2.0


#Floor Division (//): Divides the first value by the second value and returns the largest integer less than or equal to the result.
#code
result = a // b  # result is 2


#Modulus (%): Returns the remainder of the division.
#code
result = a % b  # result is 0


#Exponentiation (**): Raises the first value to the power of the second value.
#code
result = a ** b  # result is 100000


#2. Comparison Operators
#Comparison operators are used to compare two values and return a boolean result (True or False).
#Equal to (==): Checks if two values are equal.
#code
a == b  # False


#Not equal to (!=): Checks if two values are not equal.
#code
a != b  # True


#Greater than (>): Checks if the first value is greater than the second value.
#code
a > b  # True


#Less than (<): Checks if the first value is less than the second value.
#code
a < b  # False


#Greater than or equal to (>=): Checks if the first value is greater than or equal to the second value.
#code
a >= b  # True


#Less than or equal to (<=): Checks if the first value is less than or equal to the second value.
#code
a <= b  # False


#3. Logical Operators
#Logical operators are used to combine conditional statements.
#AND (and): Returns True if both conditions are true.
#code
True and False  # False


#OR (or): Returns True if at least one condition is true.
#code
True or False  # True


#NOT (not): Returns True if the condition is false.
#code
not True  # False


#4. Assignment Operators
#Assignment operators are used to assign values to variables.
#Assignment (=): Assigns a value to a variable.
#code
x = 10


#Addition Assignment (+=): Adds the right operand to the left operand and assigns the result to the left operand.
#code
x += 5  # x is now 15


#Subtraction Assignment (-=): Subtracts the right operand from the left operand and assigns the result to the left operand.
#code
x -= 3  # x is now 12


#Multiplication Assignment (*=): Multiplies the left operand by the right operand and assigns the result to the left operand.
#code
x *= 2  # x is now 24


#Division Assignment (/=): Divides the left operand by the right operand and assigns the result to the left operand.
#code
x /= 4  # x is now 6.0


#Floor Division Assignment (//=): Performs floor division and assigns the result.
#code
x //= 2  # x is now 3


#Modulus Assignment (%=): Computes the modulus and assigns the result.
#code
x %= 2  # x is now 1


#Exponentiation Assignment (**=): Raises the left operand to the power of the right operand and assigns the result.
#code
x **= 3  # x is now 1


#5. Bitwise Operators
#Bitwise operators perform operations on bits of integer types.
#AND (&): Performs a bitwise AND operation.
#code
5 & 3  # Output: 1 (binary 0101 & 0011 = 0001)


#OR (|): Performs a bitwise OR operation.
#code
5 | 3  # Output: 7 (binary 0101 | 0011 = 0111)


#XOR (^): Performs a bitwise XOR operation.
#code
5 ^ 3  # Output: 6 (binary 0101 ^ 0011 = 0110)


#Complement (~): Inverts all the bits of the operand.
#code
~5  # Output: -6 (binary ~0101 = ...1010 in two's complement)


#Left Shift (<<): Shifts the bits of the first operand to the left.
#code
5 << 1  # Output: 10 (binary 0101 << 1 = 1010)


#Right Shift (>>): Shifts the bits of the first operand to the right.
#code
5 >> 1  # Output: 2 (binary 0101 >> 1 = 0010)


#6. Membership Operators
#Membership operators test for membership within a sequence (like lists, tuples, or strings).
#In (in): Returns True if the value is found in the sequence.
#code
2 in [1, 2, 3]  # True


#Not In (not in): Returns True if the value is not found in the sequence.
#code
4 not in [1, 2, 3]  # True


#7. Identity Operators
#Identity operators compare the memory locations of two objects.
#Is (is): Returns True if both operands refer to the same object in memory.
#code
a = [1, 2, 3]
b = a
a is b  # True


#Is Not (is not): Returns True if both operands do not refer to the same object in memory.
#code
a = [1, 2, 3]
b = [1, 2, 3]
a is not b  # True


#Examples:
#Here’s how you might use these operators in a Python program:
#code
# Arithmetic Operators
x = 10
y = 3
print(x + y)  # 13
print(x - y)  # 7
print(x * y)  # 30
print(x / y)  # 3.333...
print(x // y) # 3
print(x % y)  # 1
print(x ** y) # 1000

# Comparison Operators
print(x == y) # False
print(x != y) # True
print(x > y)  # True
print(x < y)  # False
print(x >= y) # True
print(x <= y) # False

# Logical Operators
print(x > 5 and y < 5) # True
print(x > 5 or y > 5)  # True
print(not(x > 5))      # False

# Assignment Operators
x = 10
x += 5  # x is now 15
x -= 3  # x is now 12
x *= 2  # x is now 24
x /= 4  # x is now 6.0
x //= 2 # x is now 3
x %= 2  # x is now 1
x **= 3 # x is now 1

# Bitwise Operators
print(5 & 3)  # 1
print(5 | 3)  # 7
print(5 ^ 3)  # 6
print(~5)     # -6
print(5 << 1) # 10
print(5 >> 1) # 2

# Membership Operators
print(2 in [1, 2, 3])  # True
print(4 not in [1, 2, 3]) # True

# Identity Operators
a = [1, 2, 3]
b = a
print(a is b)     # True
print(a is not b) # False

#Understanding these operators and their applications is essential for writing effective Python code and solving various programming problems.
```

```python
# Type casting, also known as type conversion, is the process of converting a variable from one data type to another. 
# In Python, this can be done explicitly using built-in functions. Type casting is useful when you need to perform operations 
# that require variables to be of specific types or when you want to ensure compatibility with different types of data.

# Built-in Functions for Type Casting
# Here are some of the commonly used functions for type casting in Python:
# - int() - Converts a value to an integer.
# - float() - Converts a value to a floating-point number.
# - str() - Converts a value to a string.
# - bool() - Converts a value to a boolean.

# Examples

# 1. Converting to Integer (int())
# The int() function can convert a number or a string representation of a number to an integer.

# From Float to Integer:
x = 3.14
y = int(x)  # y is 3
print(y)    # Output: 3

# From String to Integer:
s = "123"
num = int(s)  # num is 123
print(num)    # Output: 123

# Invalid Conversion:
# This will raise a ValueError: invalid literal for int() with base 10: 'hello'
try:
    s = "hello"
    num = int(s)
except ValueError as e:
    print(e)  # Output: invalid literal for int() with base 10: 'hello'

# 2. Converting to Float (float())
# The float() function can convert a number or a string representation of a number to a floating-point number.

# From Integer to Float:
x = 5
y = float(x)  # y is 5.0
print(y)    # Output: 5.0

# From String to Float:
s = "3.14"
num = float(s)  # num is 3.14
print(num)    # Output: 3.14

# Invalid Conversion:
# This will raise a ValueError: could not convert string to float: 'abc'
try:
    s = "abc"
    num = float(s)
except ValueError as e:
    print(e)  # Output: could not convert string to float: 'abc'

# 3. Converting to String (str())
# The str() function can convert any data type to a string.

# From Integer to String:
x = 10
s = str(x)  # s is "10"
print(s)    # Output: "10"

# From Float to String:
x = 2.718
s = str(x)  # s is "2.718"
print(s)    # Output: "2.718"

# From Boolean to String:
x = True
s = str(x)  # s is "True"
print(s)    # Output: "True"

# 4. Converting to Boolean (bool())
# The bool() function converts a value to a boolean. In Python, most values are considered True except for a few cases.

# From Integer to Boolean:
x = 0
y = bool(x)  # y is False
print(y)    # Output: False

x = 1
y = bool(x)  # y is True
print(y)    # Output: True

# From String to Boolean:
x = ""
y = bool(x)  # y is False
print(y)    # Output: False

x = "hello"
y = bool(x)  # y is True
print(y)    # Output: True

# From List to Boolean:
x = []
y = bool(x)  # y is False
print(y)    # Output: False

x = [1, 2, 3]
y = bool(x)  # y is True
print(y)    # Output: True

# Type Casting in Context
# Type casting is often used in situations where you need to ensure data compatibility for operations or functions.

# Combining Data Types in Operations:
a = "5"
b = 2
result = int(a) + b  # result is 7
print(result)    # Output: 7

# Formatting Output:
age = 30
message = "I am " + str(age) + " years old."  # message is "I am 30 years old."
print(message)    # Output: "I am 30 years old."

# Understanding type casting helps avoid errors and ensures that your program handles data types correctly,
# especially when performing arithmetic operations, concatenations, or interacting with user input.
```

```python
# In Python, loops are used to execute a block of code repeatedly based on certain conditions.
# Python provides two main types of loops: for loops and while loops. Here’s a detailed overview of each type and their use cases:

# 1. for Loop
# The for loop is used to iterate over a sequence (such as a list, tuple, string, or range). It’s commonly used when you know beforehand how many times you need to iterate.

# Syntax
# for variable in sequence:
#     # Code to execute for each item in the sequence

# Use Cases
# - Iterating over elements in a list or tuple.
# - Iterating over characters in a string.
# - Generating a sequence of numbers.

# Examples

# Example 1: Iterating Over a List
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)  # Properly indented

# Output:
# apple
# banana
# cherry

# Example 2: Iterating Over a String
word = "hello"
for char in word:
    print(char)

# Output:
# h
# e
# l
# l
# o

# Example 3: Using range() Function
# The range() function generates a sequence of numbers. It’s often used with for loops to iterate a specific number of times.
for i in range(5):
    print(i)

# Output:
# 0
# 1
# 2
# 3
# 4

# Example 4: Iterating Over a Dictionary
# You can iterate over dictionary keys, values, or key-value pairs.
ages = {"Alice": 25, "Bob": 30, "Charlie": 35}
for name, age in ages.items():
    print(f"{name} is {age} years old.")

# Output:
# Alice is 25 years old.
# Bob is 30 years old.
# Charlie is 35 years old.

# 2. while Loop
# The while loop is used to execute a block of code as long as a specified condition is true. It’s commonly used when the number of iterations is not known beforehand and depends on some condition.

# Syntax
# while condition:
#     # Code to execute while the condition is true

# Use Cases
# - Repeating an action until a specific condition changes.
# - Waiting for user input until it meets certain criteria.
# - Implementing algorithms that require conditional iteration.

# Examples

# Example 1: Basic while Loop
count = 0
while count < 5:  # This is the correct while loop condition
    count += 1
    print("Count is:", count)

# This line would execute after the loop completes
print("Final count is:", count)

# Example 2: Infinite Loop with a Break Condition
while True:  # Start of the while loop
    user_input = input("Enter 'exit' to quit: ")  # This line is indented, part of the while loop block
    if user_input == "exit":  # Start of the if statement within the while loop
        break  # This line is indented, part of the if statement block
    print(f"You entered: {user_input}")  # This line is indented, part of the while loop block

# Example 3: Using while to Wait for User Input
password = ""
while password != "secret":
    password = input("Enter password: ")
    if password != "secret":
        print("Incorrect password. Try again.")
print("Password accepted!")

# Output:
# Enter password: wrong
# Incorrect password. Try again.
# Enter password: secret
# Password accepted!

# Summary
# for Loop: Ideal for iterating over sequences when the number of iterations is known or can be determined by the sequence. It’s concise and easy to use for this purpose.
# while Loop: Suitable for situations where the number of iterations is not predetermined and depends on some condition. It’s more flexible but requires careful handling to avoid infinite loops.
# Understanding these loops and their appropriate use cases allows you to write efficient and effective Python code that can handle a wide variety of tasks and scenarios.
```
