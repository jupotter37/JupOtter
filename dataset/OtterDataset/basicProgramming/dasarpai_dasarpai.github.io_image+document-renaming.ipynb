# Removing -,_, etc characters

```python
# Function to concatenate values
def remove_dashes(row):
    
    return row['Book'].replace("-_-","-").replace("-_","-").replace("_","-").replace('--','-')

# Apply the function to create the new column Z
df['New_Filename'] = df.apply(lambda row: remove_dashes(row), axis=1)
```

# PDF Page Count

```python
# !pip install PyMuPDF
```

```python
import fitz  # PyMuPDF
def count_pdf_pages(file_path):
    pdf_document = fitz.open(file_path)
    num_pages = pdf_document.page_count
    pdf_document.close()
    return num_pages

# # Replace 'your_pdf_file.pdf' with the path to your PDF file
# pdf_file_path = r'D:\02-Cloud\Hari15Sep-gdrive\DS-Res\DS-Books\Zeno-s-paradoxes-NOTE.pdf'
# page_count = count_pdf_pages(pdf_file_path)
# print(f"Number of pages in '{pdf_file_path}': {page_count}")
```

```python
import os
pref = r'D:\02-Cloud\Hari15Sep-gdrive\DS-Res\DS-Books'
for i in range( df.shape[0]):
  pdf_file_path = pref +'\\' +df.loc[i,"Book"] 
  print(i, df.loc[i,"Book"])
  size = round(os.path.getsize(pdf_file_path)/(1024*1024), 3)
  try:
    page_count = count_pdf_pages(pdf_file_path)
  except:
    print("Could not fine for file: ", pdf_file_path)
    page_count=0
  finally:
    df.loc[i,'Pages'] = page_count
    df.loc[i,'Size(MB)'] = size
    # print(page_count)
  # break
```
