```python
import requests
from bs4 import BeautifulSoup
def fetch_and_parse_text(url):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Successfully fetched the content from {url}")
        # Parse the response content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
#         print(soup)
#         Extract and print all text content from the page
        page_text = soup.get_text()
        print("Extracted Text Content:")
        print(page_text)
    else:
        print(f"Failed to fetch the content from {url}. Status code: {response.status_code}")
# Example usage
url = "https://www.python.org/"
fetch_and_parse_text(url)
```

```python
from bs4 import BeautifulSoup
def fetch_and_parse_links(url):
    # Make a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        print(f"Successfully fetched the content from {url}")
        # Parse the response content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract and print all links (URLs) from the page
        links = soup.find_all('a', href=True)
        print("Extracted Links:")
        for link in links:
            print(link['href'])
    else:
        print(f"Failed to fetch the content from {url}. Status code: {response.status_code}")
# Example usage
url = "https://www.python.org/"
fetch_and_parse_links(url)
```
