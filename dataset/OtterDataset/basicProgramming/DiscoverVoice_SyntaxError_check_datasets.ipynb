```python
import requests

# Streaming 데이터셋에서 순차적으로 항목을 가져오기
for idx, item in enumerate(python_dataset):
    if idx>10:
        break
    repo_name = item.get('repo_name')
    file_path = item.get('path')
    branch_name = item.get('branch_name', 'main')  # branch_name 필드, 기본적으로 'main' 사용
    revision_id = item.get('revision_id')  # 특정 리비전 ID가 있을 경우 활용 가능
    extension = item.get('extension')  # 파일 확장자 필드 확인
    
    
    # 경로가 제대로 있는지 확인
    if repo_name and file_path:
        # raw.githubusercontent.com URL 생성 (브랜치 고려)
        if revision_id:
            # 특정 리비전이 있는 경우에는 해당 리비전 ID를 사용
            github_url = f"https://raw.githubusercontent.com/{repo_name}/{revision_id}/{file_path}"
        else:
            # 특정 리비전이 없으면 기본적으로 브랜치 사용
            github_url = f"https://raw.githubusercontent.com/{repo_name}/{branch_name}/{file_path}"
        
        print(f"소스 코드를 가져오는 중: {github_url}")
        
        # requests를 사용해 파일 내용 다운로드 시도
        response = requests.get(github_url)
        if response.status_code == 200:
            print("소스 코드 내용:")
            print(response.text)  # 성공적으로 소스 코드를 가져온 경우 출력
        else:
            print(f"파일을 가져오지 못했습니다. 상태 코드: {response.status_code}")
            continue  # 404 등 오류 발생 시 다음 항목으로 이동
    else:
        print("레포지토리 또는 파일 경로 정보가 없습니다.")
        continue
```

```python
import requests

# Streaming 데이터셋에서 순차적으로 항목을 가져오기
for item in python_dataset:
    repo_name = item.get('repo_name')
    file_path = item.get('path')
    branch_name = item.get('branch_name', 'main')  # branch_name 필드, 기본적으로 'main' 사용
    revision_id = item.get('revision_id')  # 특정 리비전 ID가 있을 경우 활용 가능
    extension = item.get('extension')  # 파일 확장자 필드 확인
    
    # 경로가 제대로 있는지 확인
    if repo_name and file_path and extension == 'py':  # 확장자가 파이썬 파일일 때만 진행
        # raw.githubusercontent.com URL 생성 (브랜치 고려)
        if revision_id:
            # 특정 리비전이 있는 경우에는 해당 리비전 ID를 사용
            github_url = f"https://raw.githubusercontent.com/{repo_name}/{revision_id}/{file_path}"
        else:
            # 특정 리비전이 없으면 기본적으로 브랜치 사용
            github_url = f"https://raw.githubusercontent.com/{repo_name}/{branch_name}/{file_path}"
        
        print(f"소스 코드를 가져오는 중: {github_url}")
        
        # requests를 사용해 파일 내용 다운로드 시도
        response = requests.get(github_url)
        if response.status_code == 200:
            source_code = response.text  # 성공적으로 소스 코드를 가져옴
            function_lines = [line for line in source_code.splitlines() if line.strip().startswith("def ")]  # 'def '로 시작하는 줄 찾기
            if function_lines:
                print("함수 정의가 포함된 소스 코드 전체:")
                print(source_code)
                break  # 함수 정의가 있는 소스 코드를 가져오면 반복 종료
            else:
                print("함수 정의가 없습니다. 다음 항목을 확인합니다.")
                continue  # 함수 정의가 없는 경우 다음 항목으로 이동
        else:
            print(f"파일을 가져오지 못했습니다. 상태 코드: {response.status_code}")
            continue  # 404 등 오류 발생 시 다음 항목으로 이동
    else:
        print("레포지토리 또는 파일 경로 정보가 없거나 확장자가 .py가 아닙니다.")
        continue
```
