### Courses published in the year 2015

```python
udm_df.head(2)
```

```python
# Spliting the 'published_timestamp' into date and time columns
udm_df[['pub_date', 'pub_time']] = udm_df['published_timestamp'].str.split('T', expand = True)
```

```python
udm_df.head(2)
```

```python
udm_df.dtypes
```

```python
# conversion of 'pub_date' column into datetime data-type
udm_df['pub_date'] = pd.to_datetime(udm_df['pub_date'])
```

```python
udm_df.dtypes
```

```python
# extraction of 'pub_year' column from the 'pub_date' column
udm_df['pub_year'] = udm_df['pub_date'].dt.year
```

```python
udm_df.head(2)
```

```python
# courses published in 2015
udm_df[udm_df['pub_year'] == 2015].course_title.count
```
