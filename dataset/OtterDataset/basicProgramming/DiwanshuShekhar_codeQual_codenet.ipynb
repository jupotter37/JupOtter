```python
import os
import pandas as pd
py_800_dir = '../data/CodeNet/python_800'

def read_file(file):
    with open(file, 'r') as f:
        return f.read()

data = []
for root, dirs, files in os.walk(py_800_dir):
    for file in files:
        if file.endswith(".py"):
            data.append({'problem_id': root.split('/')[-1], 'submission_id': file.strip('.py'), 'submission': read_file(os.path.join(root, file))})
py800_df = pd.DataFrame(data)
py800_df
```

```python
import pandas as pd
import os

folder = '../data/CodeNet/metadata'
dataframes = []  # list to store all dataframes

for filename in os.listdir(folder):
    if filename.endswith('.csv') and filename != 'problem_list.csv':
        df = pd.read_csv(os.path.join(folder, filename))
        dataframes.append(df)

# concatenate all dataframes
metadata_df = pd.concat(dataframes, ignore_index=True)
py800_metadata = metadata_df[metadata_df['submission_id'].isin(py800_df['submission_id'])]
py800_metadata
```

```python
# group by problem_id and get the number of submissions for each problem
problem_submissions = py800_metadata.groupby('problem_id').size().reset_index(name='submissions')
problem_submissions
```

```python
# group the above dataframe by problem_id and count of number of wrong and correct submissions
grouped_df = py800_metadata.groupby('problem_id')['status'].value_counts().unstack().fillna(0)
grouped_df
```

```python
# read problem_list.csv and make a df
problem_list_df = pd.read_csv(os.path.join(folder, 'problem_list.csv'))
# rename column id to problem_id
problem_list_df.rename(columns={'id': 'problem_id'}, inplace=True)
```

```python
# enrich py800_metadata with problem_list_df['name']
py800_metadata = py800_metadata.merge(problem_list_df, on='problem_id')
py800_metadata
```

```python
from bs4 import BeautifulSoup
import os

def extract_description(content, dataset_name):
    soup = BeautifulSoup(content, 'html.parser')
    if dataset_name == 'AIZU':
        first_p = soup.find('p')
        description = first_p.text.strip() if first_p else 'NA'
        return description
    elif dataset_name == 'AtCoder':
        first_section = soup.find('section')
        if first_section:
            h3 = first_section.find('h3')
            if h3:
                h3.extract()
            description = first_section.text.strip() if first_section.text else 'NA'
            return description
        
def get_dataset_name(problem_id):
    # gets a row from problem_list_df with the given problem_id
    row = problem_list_df[problem_list_df['problem_id'] == problem_id]
    if not row.empty:
        return row['dataset'].values[0]
            

folder = '../data/CodeNet/problem_descriptions'
contents = []

for filename in os.listdir(folder):
    if filename.endswith('.html'):
        with open(os.path.join(folder, filename), 'r') as f:
            content = f.read()
            dataset_name = get_dataset_name(filename.strip('.html'))
            description = extract_description(content, dataset_name)
            contents.append({"problem_id": filename.strip('.html'), "problem": description})

# create description_df
description_df = pd.DataFrame(contents)
description_df

# merge py800_metadata with description_df
py800_metadata = py800_metadata.merge(description_df, on='problem_id')
py800_metadata
```

```python
# group by py800_metadata by dataset and get the number of problems in each dataset
problems_in_dataset = py800_metadata.groupby('dataset')['problem_id'].nunique().reset_index(name='problems')
problems_in_dataset
```

```python
# randomly sample 10 problems from each dataset
sampled_problems = py800_metadata.groupby('dataset').apply(lambda x: x.sample(n=10)).reset_index(drop=True)
sampled_problems
```

```python
# save py800_metadata to a csv file
py800_metadata.to_json('../data/CodeQualData/py800_metadata.jsonl', orient='records', lines=True)
```

```python
# from py800_df, randomly sample 15 submissions from each problem
sampled_submissions = py800_df.groupby('problem_id').apply(lambda x: x.sample(n=2)).reset_index(drop=True)
# for each row, copy files from py_800_dir to a new directory called py800_sampled
for index, row in sampled_submissions.iterrows():
    problem_id = row['problem_id']
    submission_id = row['submission_id']
    src = os.path.join(py_800_dir, problem_id, submission_id + '.py')
    dst = os.path.join('../data/CodeQualData/py800_sampled_2each', problem_id, submission_id + '.py')
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    os.system(f'cp {src} {dst}')
```

```python
# for each submission in the ../data/CodeQualData/py800_sampled directory, get the number of lines of code
data = []
for root, dirs, files in os.walk('../data/CodeQualData/py800_sampled'):
    for file in files:
        if file.endswith(".py"):
            with open(os.path.join(root, file), 'r') as f:
                lines = f.readlines()
                data.append({'problem_id': root.split('/')[-1], 'submission_id': file.strip('.py'), 'lines_of_code': len(lines)})
            
#order by descending order of lines_of_code
lines_of_code_df = pd.DataFrame(data)
lines_of_code_df = lines_of_code_df.sort_values(by='lines_of_code', ascending=False)
lines_of_code_df
```

```python
# for each submission in the ../data/CodeQualData/py800_sampled directory, get the size of the file
data = []
for root, dirs, files in os.walk('../data/CodeQualData/py800_sampled'):
    for file in files:
        if file.endswith(".py"):
            data.append({'problem_id': root.split('/')[-1], 'submission_id': file.strip('.py'), 'size': os.path.getsize(os.path.join(root, file))})

#order by descending order of size
size_df = pd.DataFrame(data)
size_df = size_df.sort_values(by='size', ascending=False)
size_df
```
