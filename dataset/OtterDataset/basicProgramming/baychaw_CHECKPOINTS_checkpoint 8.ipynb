```python
import requests
from bs4 import BeautifulSoup

def get_wikipedia_text(page_title):
    # URL de la page Wikipédia
    url = f"https://en.wikipedia.org/wiki/{page_title}"

    # Faire la requête pour récupérer le contenu HTML de la page
    response = requests.get(url)

    # Vérifier si la requête a réussi (code 200)
    if response.status_code == 200:
        # Utiliser BeautifulSoup pour analyser le HTML
        soup = BeautifulSoup(response.text, 'html.parser')

        # Trouver le contenu du corps de la page
        body_content = soup.find('div', {'id': 'bodyContent'})

        # Extraire le texte du contenu du corps
        text_content = body_content.get_text(separator='\n')

        return text_content
    else:
        # Afficher un message d'erreur si la requête a échoué
        print(f"Erreur {response.status_code} lors de la récupération de la page.")
        return None

# Exemple d'utilisation
page_title = "Python_(programming_language)"
wikipedia_text = get_wikipedia_text(page_title)

# Afficher le texte (si disponible)
if wikipedia_text:
    print(wikipedia_text)
```
