## Linting Python Scripts

The goals of this session are to:

* Help to standardise python scripts prior to sharing online or with team members.
* Introduce pylint as a tool used to standardise this process.

Linting your code involves running software designed to identify bugs, improve readability and performance.

***

## Why do we lint?

It is recommended to lint code prior to sharing scripts with colleagues or online. There are various approaches to linting code. In this activity, we will be exploring the use of **Pylint**. Some companies such as [Google](https://google.github.io/styleguide/pyguide.html) incorporate linting software as part of their programming style guide. 

***
   
## How do I lint with Pylint?

<img src="https://cdn-learn.adafruit.com/assets/assets/000/079/697/medium260/circuitpython_2000px-Pylint_Logo_2018.svg.png?1566404053"
         alt="Image from [learn.adafruit.com/](https://learn.adafruit.com/improve-your-code-with-pylint?view=all)"
         width="400"
         style="margin: 0px 60ps"
         />
         
If you use a distribution of **Anaconda**, it is likely that you already have Pylint installed. The [user manual](http://pylint.pycqa.org/en/latest/) is extensive and contains all the information you will need. 

Pylint can be run from **Anaconda prompt** on Windows or **Terminal** on Mac. From your console type:

``` Pylint path/to/python_script.py ```

If you have Pylint installed then the program will execute within the console window. If you need to install Pylint, you can do so using pip:

``` pip install pylint ```

Once you have executed pylint over a script, the linting software will identify issues within your script and output them to your console. Let's look at a portion of this output together.

<img src="images/Screenshot 2020-08-18 at 08.48.46.png"
         alt="Sample Pylint output"
         width="800"
         style="margin: 0px 60ps"
         />








This output can look a little overwhelming if you're a messy coder (like me). The idenitfied issues are quite readable, though. For example, by opening up the titanic_EDA.py script and navigating to **line 6**, you will see that the line is flagged as too long. By spacing this over two lines, the readability of the script is improved. 

***

Another feature of Pylint to note is the scoring system.

<img src="images/Screenshot 2020-08-18 at 08.49.55.png"
         alt="Pylint script scoring"
         width="800"
         style="margin: 0px 60ps"
         />


At the bottom of the pylint output you can see a score out of 10. Yes, it's possible to score negative marks!

If you ensure you keep your Terminal or Anaconda Prompt console open between successive pylint runs, it also shows you your improvement on your previous pylint call.
