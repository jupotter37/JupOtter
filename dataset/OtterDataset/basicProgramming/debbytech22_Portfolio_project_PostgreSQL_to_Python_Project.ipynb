# The aim of the project is to create a connection between postgreSQL and python in order to query the dataset from python environment 

```python
#import all the necessary libraries for python and sql connection
import psycopg2
import psycopg2.extras
import pandas as pd
from sqlalchemy import create_engine
import requests
```

```python
df=pd.read_csv('Untitled spreadsheet - udemy_courses.csv')
```

```python
#create a connection between postgre and python
connection=psycopg2.connect(
    host='localhost',
    database='demo_db',
    user='postgres',
    password='postgres'                                                   
)
```

```python
pointer=connection.cursor()
```

```python
# A query for paid courses, showing the course with the highest sales per month
qry='''select course_title,EXTRACT(MONTH FROM published_timestamp) AS months,
count(distinct (is_paid) ) as high
from udemy_courses 
group by course_title,months'''
record=pointer.execute(qry)
pointer.fetchall()
```

```python
# A query for paid courses, showing the  course with the highest number of subscribers per month
query='''select course_title,EXTRACT(MONTH FROM published_timestamp) AS months,
max(num_subscribers) as high
from udemy_courses 
group by course_title,months'''
record=pointer.execute(query)
pointer.fetchall()
```
