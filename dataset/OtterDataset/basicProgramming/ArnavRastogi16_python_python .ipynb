 ## QUES 5) Explain the concept of type casting in Python with examples .
 
 In Python, type casting refers to the conversion of one data type to another. This process allows you to change the data type of a variable or value. Here's an explanation of type casting with examples:

1.*Implicit Type Conversion*:
   -Python automatically converts data types in certain situations, such as when performing operations between different types.
   
   Example:
   python
   
   num_int = 10
   
   num_float = 5.5
   
   result = num_int + num_float  # Python implicitly converts num_int to a float before performing the addition
   print(result)  # Output: 15.5
   

2.*Explicit Type Conversion*:
   -You can explicitly convert data types using predefined functions like int(), float(), str(), etc.
   
   Example:
   python
   
   num_str = "50"
   
   num_int = int(num_str)  # Convert the string "50" to an integer
   print(num_int)  # Output: 50

   num_float = float(num_int)  # Convert the integer 50 to a float
   print(num_float)  # Output: 50.0
   

3.*Type Casting between Different Data Types*:
   -You can convert data types between integers, floats, and strings.
   
   Example:
   python
   
   num_int = 100
   
   num_str = str(num_int)  # Convert the integer 100 to a string
   print(num_str)  # Output: "100"

   num_float = float(num_str)  # Convert the string "100" to a float
   print(num_float)  # Output: 100.0
   

4.*Handling Errors during Type Casting*:
   -Type casting can raise errors if the conversion is not possible (e.g., converting a string that is not a number to an integer).
   
   Example:
   python
   
   num_str = "hello"
   
   try:
   
num_int = int(num_str)  # Trying to convert a non-numeric string to an integer will raise a ValueError
   except ValueError as e:
       print("Error:", e)
   

Type casting is a useful technique in Python to manipulate data types and perform operations that require values of different types.

## QUES 7) Describe the different types of loops in Python and their use cases with examples.

In Python, there are mainly two types of loops: for loops and while loops. Each type of loop has its use cases and can be helpful in different situations.

1.*For Loop*:
   -Use a for loop when you know the number of iterations in advance or when you want to iterate over a sequence (like a list, tuple, string, etc.).
   -Here's an example of a for loop:

python
#Example of a for loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)


In this example, the for loop iterates over each element in the fruits list and prints each fruit.

2.*While Loop*:
   -Use a while loop when you want to iterate based on a condition that needs to be true.
   -Here's an example of a while loop:

python
#Example of a while loop
count = 0
while count < 5:
    print(count)
    count += 1


In this example, the while loop will keep executing as long as the count is less than 5. It prints the value of count and increments it by 1 in each iteration.

Both for and while loops are essential in programming and help in automating repetitive tasks or processing elements in a sequence.
