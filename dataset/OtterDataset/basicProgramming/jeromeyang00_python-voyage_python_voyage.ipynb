### STRING OPERATIONS

- String operations are commonly used as methods in accessing string data type for data cleaning and debugging in some cases.



````python
# Define a sample string
sample_string = "Hello, World!"

# Length of a string
length = len(sample_string)
print(f"Length: {length}")  # Output: 13

# Accessing characters by index
first_char = sample_string[0]
last_char = sample_string[-1]
print(f"First character: {first_char}, Last character: {last_char}")  # Output: H, !

# Converting to uppercase
uppercase = sample_string.upper()
print(f"Uppercase: {uppercase}")  # Output: HELLO, WORLD!

# Converting to lowercase
lowercase = sample_string.lower()
print(f"Lowercase: {lowercase}")  # Output: hello, world!

# Capitalizing the first character
capitalized = sample_string.capitalize()
print(f"Capitalized: {capitalized}")  # Output: Hello, world!

# Title case (capitalize the first character of each word)
title_case = sample_string.title()
print(f"Title Case: {title_case}")  # Output: Hello, World!

# Stripping whitespace
whitespace_string = "   Hello, World!   "
stripped = whitespace_string.strip()
print(f"Stripped: '{stripped}'")  # Output: 'Hello, World!'

# Splitting a string into a list
split_string = sample_string.split(", ")
print(f"Split: {split_string}")  # Output: ['Hello', 'World!']

# Joining a list into a string
joined_string = ", ".join(split_string)
print(f"Joined: {joined_string}")  # Output: Hello, World!

# Checking if a string starts with a substring
starts_with = sample_string.startswith("Hello")
print(f"Starts with 'Hello': {starts_with}")  # Output: True

# Checking if a string ends with a substring
ends_with = sample_string.endswith("!")
print(f"Ends with '!': {ends_with}")  # Output: True

# Finding the index of a substring
index_of_world = sample_string.find("World")
print(f"Index of 'World': {index_of_world}")  # Output: 7

# Counting occurrences of a substring
count_of_l = sample_string.count("l")
print(f"Count of 'l': {count_of_l}")  # Output: 3

# String formatting
name = "Alice"
age = 30
formatted_string = f"Name: {name}, Age: {age}"
print(f"Formatted String: {formatted_string}")  # Output: Name: Alice, Age: 30

# Escaping special characters
escaped_string = "He said, \"Hello, World!\""
print(f"Escaped String: {escaped_string}")  # Output: He said, "Hello, World!"

# Reversing a string (not a direct method, but can be done with slicing)
reversed_string = sample_string[::-1]
print(f"Reversed String: {reversed_string}")  # Output: !dlroW ,olleH

````

   - Defined using square brackets `[]`.

2. **Tuples**:

- **Lists**: Indexing, slicing, appending, removing, modifying elements.
- **Tuples**: Indexing, returning multiple values from functions, immutable.
- **Sets**: Adding, removing elements, handling duplicates, set operations (union, intersection, difference).

````python
# Lists
# Lists are ordered, mutable sequences of elements. They can contain elements of different types.
# Lists are defined using square brackets [].

# Creating a list
fruits = ["apple", "banana", "cherry"]
print(f"List: {fruits}, Type: {type(fruits)}")

# Accessing elements by index
print(f"First fruit: {fruits[0]}")  # Output: apple
print(f"Last fruit: {fruits[-1]}")  # Output: cherry

# Modifying elements
fruits[1] = "blueberry"
print(f"Modified list: {fruits}")  # Output: ['apple', 'blueberry', 'cherry']

# Adding elements
fruits.append("date")
print(f"Appended list: {fruits}")  # Output: ['apple', 'blueberry', 'cherry', 'date']

# Removing elements
fruits.remove("apple")
print(f"List after removal: {fruits}")  # Output: ['blueberry', 'cherry', 'date']

# Tuples
# Tuples are ordered, immutable sequences of elements. They can contain elements of different types.
# Tuples are defined using parentheses ().

# Creating a tuple
coordinates = (10.0, 20.0)
print(f"Tuple: {coordinates}, Type: {type(coordinates)}")

# Accessing elements by index
print(f"First coordinate: {coordinates[0]}")  # Output: 10.0
print(f"Second coordinate: {coordinates[1]}")  # Output: 20.0

# Tuples are immutable, so elements cannot be modified
# coordinates[0] = 15.0  # This line would raise an error


# Sets
# Sets are unordered collections of unique elements. They cannot contain duplicate elements.
# Sets are defined using curly braces {} or the set() function.

# Creating a set
unique_fruits = {"apple", "banana", "cherry"}
print(f"Set: {unique_fruits}, Type: {type(unique_fruits)}")

# Adding elements
unique_fruits.add("date")
print(f"Set after adding: {unique_fruits}")  # Output: {'apple', 'banana', 'cherry', 'date'}

# Removing elements
unique_fruits.remove("banana")
print(f"Set after removal: {unique_fruits}")  # Output: {'apple', 'cherry', 'date'}

# Sets automatically handle duplicates
unique_fruits.add("apple")
print(f"Set after adding duplicate: {unique_fruits}")  # Output: {'apple', 'cherry', 'date'}

# Set operations
another_set = {"date", "elderberry", "fig"}

# Union: elements in either set
union_set = unique_fruits.union(another_set)
print(f"Union: {union_set}")  # Output: {'apple', 'date', 'cherry', 'elderberry', 'fig'}

# Intersection: elements in both sets
intersection_set = unique_fruits.intersection(another_set)
print(f"Intersection: {intersection_set}")  # Output: {'date'}

# Difference: elements in the first set but not the second
difference_set = unique_fruits.difference(another_set)
print(f"Difference: {difference_set}")  # Output: {'apple', 'cherry'}
````

## Dictionaries
- Dictionaries offer efficient and flexible data storage, making them indispensable in many programming tasks.
- Understanding keys and values is essential for working effectively with dictionaries.
- Printing just keys or values can be useful for specific operations, such as iterating or accessing specific data.

## Importance of dictionaries
- Dictionaries allow efficient retrieval of values using unique keys.
- They are versatile data structures used for mappings, caching, and representing structured data.
- Dictionaries are commonly used in various domains, including web development, data processing, and system programming.


````python
# Creating a dictionary with student names and their corresponding grades
student_grades = {
    "Alice": 85,
    "Bob": 70,
    "Charlie": 95,
    "David": 60,
    "Emma": 75
}

# Printing the dictionary
print("Dictionary of student grades:", student_grades)

# Accessing values using keys
alice_grade = student_grades["Alice"]
print("\nAlice's grade:", alice_grade)

# Modifying values
student_grades["Bob"] = 80
print("\nUpdated dictionary after modifying Bob's grade:", student_grades)

# Adding new key-value pairs
student_grades["Frank"] = 90
print("\nUpdated dictionary after adding Frank's grade:", student_grades)

# Removing key-value pairs
removed_grade = student_grades.pop("David")
print("\nRemoved grade of David:", removed_grade)
print("Updated dictionary after removing David's grade:", student_grades)

# Printing just the keys
print("\nJust the keys:")
for name in student_grades.keys():
    print(name)

# Printing just the values
print("\nJust the values:")
for grade in student_grades.values():
    print(grade)

# Explanation of keys and values
print("\nKeys represent student names, and values represent their grades.")
print("Keys:", student_grades.keys())
print("Values:", student_grades.values())
````

# Flow Control in Programming

Flow control refers to the order in which statements are executed in a program. It allows programmers to dictate the path a program takes based on conditions, loops, and function calls.

## Types of Flow Control:

1. **Conditional Execution**: This involves using conditions (e.g., if statements) to execute specific blocks of code based on whether certain conditions are met.

2. **Looping**: Looping structures (e.g., for loops, while loops) allow repetitive execution of code blocks until a condition is satisfied.

3. **Function Calls**: Functions are blocks of reusable code that can be called upon to perform specific tasks, allowing for modular and structured programming.

Flow control is essential for creating programs that can make decisions, repeat tasks, and organize code efficiently.

# Looping in Flow Control

Looping is a fundamental concept in programming that allows the execution of a block of code repeatedly. It helps automate repetitive tasks and iterate over data structures. There are several types of loops in Python:

1. **for Loop**: Executes a block of code iteratively for a predefined number of times or over an iterable object like lists, tuples, strings, or dictionaries.
2. **while Loop**: Executes a block of code repeatedly as long as a specified condition is true.
3. **Nested Loops**: You can nest loops within one another to perform complex iterations.

Loops are essential for tasks like iterating over elements in a list, processing data, or performing calculations. They provide flexibility and efficiency in handling repetitive tasks.



````python
import json

# Load the JSON data from the file
with open("data.json", 'r') as file:
    data = json.load(file)

# Accessing the list of students
students = data.get('students', [])

# Using a for loop to check if each student has failed in any course
print("Using for loop:")
for student in students:
    name = student['name']
    grades = student['grades']
    
    # Check if any grade is equal to 3 (indicating failure)
    if 3 in grades.values():
        print(f"{name} has failed in at least one course.")
    else:
        print(f"{name} has passed all courses.")
````

````python

# Using a while loop to achieve the same result
print("\nUsing while loop:")
num_students = len(students)
index = 0
while index < num_students:
    name = students[index]['name']
    grades = students[index]['grades']
    
    # Check if any grade is equal to 3 (indicating failure)
    if 3 in grades.values():
        print(f"{name} has failed in at least one course.")
    else:
        print(f"{name} has passed all courses.")
    
    index += 1
````

## Introduction to Python Functions

In Python, a function is a block of code that performs a specific task. Functions allow you to write reusable code, making your programs more modular and easier to maintain. 

### Function Definition

A function is defined using the `def` keyword, followed by the function name, parentheses containing any parameters, and a colon. The body of the function contains the code to be executed.

### Arguments and Parameters

- **Parameters** are variables listed inside the parentheses in the function definition.
- **Arguments** are the values passed to the function when it is called.

### Example Function

Here is a simple example of a Python function that takes two parameters and returns their sum:

```python
# Function definition with parameters
def add_numbers(a, b):
    """
    This function takes two numbers and returns their sum.
    """
    result = a + b
    return result

# Calling the function with arguments
sum_result = add_numbers(5, 3)
print(f"The sum is: {sum_result}")  # Expected Output: The sum is: 8
```

- **Function Definition:** `def add_numbers(a, b):` defines a function named `add_numbers` with two parameters `a` and `b`.
- **Function Body:** The code inside the function calculates the sum of `a` and `b` and stores it in the variable `result`.
- **Return Statement:** `return result` sends the result back to the caller of the function.
- **Function Call:** `add_numbers(5, 3)` calls the function with the arguments `5` and `3`.

````python
def add_numbers(a, b):
    """
    This function takes two numbers and returns their sum.
    """
    # Calculate the sum of the two numbers
    sum_result = a + b
    # Return the result
    return sum_result

# Calling the function with arguments without type hinting
a = input("A:")
b = input("B:")
sum_result = add_numbers(a,b)
print(f"The sum is: {sum_result}")  

a = int(input("A:"))
b = int(input("B:"))
sum_result = add_numbers(a,b)
print(f"The sum is: {sum_result}")  
````

# Understanding `*args` and `**kwargs` in Python

When defining functions in Python, you may want to pass a variable number of arguments. This is where `*args` and `**kwargs` come in handy.

## `*args`

- `*args` allows a function to accept any number of positional arguments.
- It is used to pass a variable number of arguments to a function.
- Inside the function, `args` is a tuple that contains all the positional arguments passed to the function.

### Example

```python
def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(1, 2, 3, 4, 5)
```

In this example, `print_numbers` can accept any number of arguments, which are then printed one by one.

## `**kwargs`

- `**kwargs` allows a function to accept any number of keyword arguments.
- It is used to pass a variable number of keyword arguments to a function.
- Inside the function, `kwargs` is a dictionary that contains all the keyword arguments passed to the function.

### Example

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="Wonderland")
```

In this example, `print_info` can accept any number of keyword arguments, which are then printed as key-value pairs.

## Combining `*args` and `**kwargs`

You can use both `*args` and `**kwargs` in the same function to accept both positional and keyword arguments.

### Example

```python
def display(*args, **kwargs):
    for arg in args:
        print(f"Arg: {arg}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display(1, 2, 3, name="Alice", age=30)
```
