```python
non_unnamed_columns = {i:col for i, col in enumerate(students_data.columns) if not col.startswith('Unnamed')}
```

```python
student_data = pd.read_csv(r'Careers_recommendation\Student_resume_data.csv',skiprows=1)
```

```python
student_data.head()
```

```python
new_column_names = {student_data.columns[idx]: new_name for idx, new_name in non_unnamed_columns.items() if idx < len(student_data.columns)}
```

```python
new_column_names
```

```python
student_data = student_data.rename(columns=new_column_names)
```

```python
student_data.head()
```

```python
students_df = student_data.loc[:, ~student_data.columns.str.startswith(('Self', 'Validation'))]
```

```python
students_df.head()
```

```python
students_df.columns
```

```python
remove_cols = ['Unnamed: 23','Graduation Percentage/CGPA','Post Graduation Percentage / CGPA','Year of Graduation Completion',
               'Year of Post Graduation Completion',
               'Graduation College','Post Graduation College',
               'Domain / Industry Interested to Join as a Data Scientist ','Company\nName','Relevant Work Exp\n(years)',
               'Post Graduation Degree','Date Of Birth\n(DD/MM/YYYY)',
               'Current Location','Weakness\n(Any including tools, skills)']
```

```python
students_df = students_df.drop(columns = remove_cols, axis = 1)
```

```python
students_df.head()
```

```python
# renaming the columns
rename_cols = {"Strength\n(Any including tools, skills)":'Strengths', 'Notice Period (days)':'NP(days)','Domain\n/Industry' :'Domain', 
               'Current/Previous CTC\nin LPA':'Current CTC' }
```

```python
students_df = students_df.rename(columns = rename_cols)
```

```python
# data info
students_df.info()
```

```python
students_df.isnull().sum()
```

```python
# removing the rows with no student name
students_df[students_df['Student Name'].isnull() ==  True]
```

```python
students_df = students_df.dropna(subset=['Student Name'])
```

```python
students_df['Post_Grad'] = students_df['Post Graduation Stream'].notnull().astype(int)
students_df = students_df.drop(columns = ['Post Graduation Stream'], axis=1)
```

```python
students_df['NP(days)'] = students_df['NP(days)'].str.extract('(\d+)')
students_df['NP(days)'] = students_df['NP(days)'].astype('float')
students_df.fillna({'NP(days)': students_df['NP(days)'].max() }, inplace = True)
```

```python
students_df['Work Exp (years)'] = students_df['Work Exp (years)'].str.extract('(\d+)')
students_df['Work Exp (years)'] = students_df['Work Exp (years)'].astype('float')
students_df['Work Exp (years)'] = students_df['Work Exp (years)'].fillna(0.0)
```

```python
students_df['Current CTC'] = students_df['Current CTC'].astype('float')
students_df.fillna({'Current CTC': 0.0},inplace=True)
```

```python
students_df['Domain'] = students_df['Domain'].fillna('None')
students_df['Designation'] = students_df['Designation'].fillna('Fresher')
students_df['AWS'] = students_df['AWS'].astype(bool)
```

```python
# converting the bool tp int format
skills = ['AWS', 'Azure', 'GCP', 'PySpark',
       'Hadoop', 'Hive', 'Big Query', 'Tableau', 'PowerBI', 'Advanced Excel',
       'Time Series Analysis', 'ML Model Deployment', 'RestAPI', 'ETL',
       'Deep Learning', 'Tensorflow', 'PyTorch']
```

```python
for col in skills:
    students_df[col] = students_df[col].astype(int)
```

```python
students_df.to_csv('data/students_data.csv')
```
