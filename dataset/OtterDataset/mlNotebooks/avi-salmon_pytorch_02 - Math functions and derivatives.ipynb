## Part 1: Understanding Functions

### What is a Function?

A **function** is a relationship between a set of inputs and a set of possible outputs where each input is related to exactly one output. In mathematical terms, if we have an input $x$, a function $f$ maps it to an output $y$:

$y = f(x)$

### Examples of Functions

* **Linear Function:** $f(x) = 2x + 3$
* **Quadratic Function:** $f(x) = x^2$
* **Sine Function:** $f(x) = \sin(x)$

### Python Functions



```python
def linear_function(x):
    return 2 * x + 3
```

## Part 2: Plotting Functions
To visualize functions, we'll use the matplotlib library.

Importing Necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create an array of x values
x_values = np.linspace(-5, 5, 100)

# Compute y values
y_values = linear_function(x_values)

# Plot the function
plt.plot(x_values, y_values, label=r'$f(x) = 2x + 3$')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Plotting a Non-Linear Function

Let's plot $f(x) = x^2$.



```python
def quadratic_function(x):
    return x ** 2

y_values = quadratic_function(x_values)

plt.plot(x_values, y_values, label=r'$f(x) = x^2$')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

# Part 3: Derivatives of Functions

## What is a Derivative?

The **derivative** of a function measures how the function value changes as its input changes. It's a fundamental tool in calculus and is essential for optimization in machine learning.

Mathematically, the derivative of $f(x)$ with respect to $x$ is denoted as $f'(x)$ or $\frac{df}{dx}$.

## Computing Derivatives

### Analytically

* For $f(x) = x^2$, the derivative is:

  $f'(x) = 2x$

* For $f(x) = \sin(x)$, the derivative is:

  $f'(x) = \cos(x)$

## Visualizing the Derivative

Let's plot the function $f(x) = x^2$ and its derivative $f'(x) = 2x$.



```python
# Original function
y_values = quadratic_function(x_values)

# Derivative function
def derivative_quadratic_function(x):
    return 2 * x

# def derivative_quadratic_function(x):
#     h = 0.00001
#     return (quadratic_function(x+h) - quadratic_function(x)) / h

y_derivative_values = derivative_quadratic_function(x_values)

# Plot both functions
plt.plot(x_values, y_values, label=r'$f(x) = x^2$')
plt.plot(x_values, y_derivative_values, label=r"$f'(x) = 2x$")
plt.title('Function and its Derivative')
plt.xlabel('x')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```
