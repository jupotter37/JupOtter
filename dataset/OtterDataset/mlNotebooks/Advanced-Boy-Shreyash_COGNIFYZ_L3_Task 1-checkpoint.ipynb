```python
import requests
from bs4 import BeautifulSoup as bs 
response = requests.get("https://cognifyz.com/")                                                       
print(response.status_code)
```

```python
scrap = bs(response.content, "html.parser")  
print(scrap)
```

```python
print(scrap.title)
```

```python
print(scrap.title.string)
```

```python
print(scrap.find_all("b"))
```

```python
print(scrap.find_all("button"))
```

```python
print(scrap.find_all("link"))
```

```python
for link in scrap.find_all('a'):
    print(link.get('href'))
```
