# Pandas

```python
import pandas as pd
df = pd.read_csv('temporal.csv')
```

```python
print(df)
df.describe() #pandas dataframe descStat függvénye!
```

```python
#Adatok megjelnítése más formátumbnan, csak vizuálisan változik
format_dict = {'data science':'{0:,.2f} HUF','date':'{:%m-%Y}', 'machine learning':'{:.2%}'}

#We make sure that the Month column has datetime
df['date'] = pd.to_datetime(df['date'])

#We apply the style to the visualization
df.head().style.format(format_dict) # a létrehozott maszkot alkalmazzuk a megjelenítés során
```

```python
#Adatok színezése
format_dict = {'date':'{:%m-%Y}'} # A maszk deklarását írjuk át csak a dátum oszlopra, így több értelme lesz a használatának.
df.head(10).style.format(format_dict).highlight_max(color='darkgreen').highlight_min(color='#ff0000')

#Színátmenetes változat
df.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'],cmap='BuGn')

#Sávos színezés
df.head(10).style.format(format_dict).bar(color='red', subset=['data science', 'deep learning'])

#Előzőek kombinálva
df.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'],cmap='BuGn').highlight_max(color='yellow')
```

# Matplotlib

```python
import matplotlib.pyplot as plt

plt.plot(df['date'], df['data science'], label='datascience')
plt.plot(df['date'], df['machine learning'],label='machine learning')
plt.plot(df['date'], df['deep learning'], label='deeplearning')
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by date')
plt.grid(True)
plt.legend()
```

```python
#Több ábra egyszerre
fig, axes = plt.subplots(2,2)
axes[0, 0].hist(df['data science'])
axes[0, 1].scatter(df['date'], df['data science'])
axes[1, 0].plot(df['date'], df['machine learning'])
axes[1, 1].plot(df['date'], df['deep learning'])
```

```python
plt.plot(df['date'], df['data science'], 'r-')
plt.plot(df['date'], df['data science']*2, 'bs')
plt.plot(df['date'], df['data science']*3, 'g^')
plt.grid(True)
```

```python
#x-y scatter plot
plt.scatter(df['data science'], df['machine learning'])
plt.xlabel('data science')
plt.ylabel('machine learning')
```

```python
#Barchart
plt.bar(df['date'], df['machine learning'], width=20)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('machine learning')
```

```python
#hisztogram
plt.hist(df['deep learning'], bins=15)
plt.title('deep learning')
```
