Q2- Supervised learning is a machine learning technique where the model is trained on labeled data. Labeled data means that each input data point is associated with a corresponding correct output or target value. The goal of supervised learning is to learn a mapping between inputs and outputs based on the provided labeled examples. Some examples of supervised learning algorithms include:

- Linear regression
- Logistic regression
- Support Vector Machines (SVM)
- Decision trees
- Random forests
- Gradient Boosting algorithms (e.g., XGBoost, LightGBM)
- Neural networks (e.g., feedforward neural networks)

Q3- Unsupervised learning is a machine learning technique where the model learns patterns and structures in the data without any explicit target or output labels. The goal of unsupervised learning is to discover inherent structures or relationships within the data. Some examples of unsupervised learning algorithms include:

- Clustering algorithms (e.g., K-means, DBSCAN, hierarchical clustering)
- Dimensionality reduction techniques (e.g., Principal Component Analysis (PCA), t-SNE)
- Association rule learning (e.g., Apriori algorithm)
- Generative models (e.g., Gaussian Mixture Models (GMM), Variational Autoencoders (VAE))
- Self-organizing maps

Q4- AI (Artificial Intelligence) is a broad field that encompasses the development of intelligent machines that can perform tasks that typically require human intelligence. Machine Learning (ML) is a subset of AI that focuses on the development of algorithms and models that enable computers to learn from data and make predictions or take actions without being explicitly programmed. Deep Learning (DL) is a subfield of ML that utilizes artificial neural networks with multiple layers to extract high-level features from data. Data Science (DS) is an interdisciplinary field that combines ML, statistical analysis, data visualization, and domain knowledge to extract insights and knowledge from data.

Q5- The main differences between supervised, unsupervised, and semi-supervised learning are:

Supervised Learning:
- Requires labeled training data with input-output pairs.
- Learns to make predictions based on labeled examples.
- Supervised learning algorithms aim to minimize the difference between predicted outputs and true outputs.
- Examples: Image classification, sentiment analysis, regression tasks.

Unsupervised Learning:
- Uses unlabeled data, without explicit input-output pairs.
- Learns patterns, structures, and relationships in the data.
- Unsupervised learning algorithms aim to discover inherent structures or groupings in the data.
- Examples: Clustering, dimensionality reduction, anomaly detection.

Semi-supervised Learning:
- Utilizes a combination of labeled and unlabeled data.
- Leverages both labeled and unlabeled examples to improve learning accuracy.
- Can be useful when obtaining labeled data is expensive or time-consuming.
- Examples: Document classification, speech recognition, protein structure prediction.

Q6- In machine learning, the training, test, and validation split refers to the division of the available data into separate subsets for different purposes:

- Training Set: This is the portion of the data used to train the machine learning model. The model learns from the patterns and relationships present in the training set.

- Test Set: This is an independent subset of the data that is used to evaluate the performance of the trained model. The test set is used to assess how well the model generalizes to new, unseen data.

- Validation Set: In some cases, a validation set is used during the training process to tune hyperparameters or make decisions about the model. The validation set helps in evaluating different models or configurations and selecting the best one.

The importance of these splits lies in evaluating the performance and generalization of the trained model. The training set is used to learn the model's parameters, the validation set helps in tuning the model, and the test set provides an unbiased assessment of the model's performance on unseen data.

Q7- Unsupervised learning can be used in anomaly detection by learning patterns and structures in the normal behavior of the data. Once the model is trained on normal data, it can identify instances that deviate significantly from the learned patterns as potential anomalies. The anomalies are detected based on the dissimilarity or low probability scores assigned to them by the unsupervised learning algorithm. This approach allows for the detection of unusual or abnormal data points without requiring labeled examples of anomalies.
