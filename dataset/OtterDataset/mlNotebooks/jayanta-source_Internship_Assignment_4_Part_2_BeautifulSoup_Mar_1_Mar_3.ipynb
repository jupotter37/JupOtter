```python
from requests_html import AsyncHTMLSession
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

Question 1: Please visit https://www.cnbc.com/world/?region=world and scrap-

a) headings
b) date
c) News link

```python
async def CNBC_news(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender(sleep=5)
    
    soup=BeautifulSoup(response.html.html,'lxml')
    
    headings = []
    published_date = []
    news_links = []
    
    headings_body = soup.find('div',class_='RiverPlus-riverPlusContainer')
    
    for heading_details in headings_body.find_all('div',class_='RiverHeadline-headline RiverHeadline-hasThumbnail'):
        headings.append(heading_details.text)
        a_tags=heading_details.find_all('a')
        if len(a_tags)>=2:
            news_links.append(a_tags[1]['href'])
        else:
            news_links.append(a_tags[0]['href'])
    
    
    pattern=r'/(\d{4})/(\d{2})/(\d{2})'
    
    
    for url in news_links:
        date=re.findall(pattern,url)
        if date:
            published_date.append('/'.join(date[0]))
        else:
            published_date.append('Date not available')
        
    
    
    df=pd.DataFrame({'News Title':headings,'Date Published':published_date,'URL Links':news_links})
    return df
        
    
    
await CNBC_news('https://www.cnbc.com/world/?region=world')
```

Questions 2: Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/ and scrap

a) Paper title
b) date
c) Author

```python
async def Keaipublishing(url):
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender(sleep=5)
    
    paper_titles = []
    date = []
    author = []
    
    soup=BeautifulSoup(response.html.html,'lxml')
    
    for article_details in soup.find_all('div',class_='article-listing'):
        title = article_details.find('h2',class_='h5 article-title')
        paper_titles.append((title.find('a')).text.strip())
        
        date.append((article_details.find('p',class_='article-date')).text)
        
        author.append((article_details.find('p',class_='article-authors')).text)
        
    df = pd.DataFrame({'Paper Title':paper_titles,'Published date':date,'Authors':author})
    return df
   


    
await Keaipublishing('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')    
```
