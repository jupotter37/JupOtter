```

import pandas as pd
import re
import numpy as np
from nltk.translate.bleu_score import sentence_bleu
pd.set_option("display.max_colwidth", 10000)

chatbot_data = pd.read_csv("./output5.csv", keep_default_na = False)

r, c = chatbot_data.shape
def calcBleuScore(actual,expected):
  score = sentence_bleu(actual, expected)
 
  return score;
  
```

```
import re
count =0
print("r ",r)
for i in range(0,r):
  question = chatbot_data.loc[i][0].lower()
  actual_answer = chatbot_data.loc[i][1].lower()
  expected_answer = chatbot_data.loc[i][2].lower()
  actual_answer = re.sub(r"([.!?,])", r" \1",actual_answer)
  actual_answer = actual_answer.strip('\"')
  expected_answer = expected_answer.strip('\"')
  expected_answer = re.sub(r"([.!?,])", r" \1", expected_answer)
  a=actual_answer.split()
  b=expected_answer.split()
  c=[]
  c.append(a)
  score = calcBleuScore(c,b)
  if score > 0.75:
   count =  count+1
  else:
    print("question ====",question)
    print("======= actual_answer   ===",actual_answer)
    print("======= expected_answer ===",expected_answer)
    

print("count",count)
percentage = (count/r)*100
print("percentage",percentage)
```
