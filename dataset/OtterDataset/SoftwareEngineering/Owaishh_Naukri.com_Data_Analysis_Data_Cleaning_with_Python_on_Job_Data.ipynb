# Naukri Data Job Posting Data Cleaning

This is the data cleaning step of a database file containing Jobs with details like Job Name,Company, rating, minimum years of experience , salary offered , date etc. 

First process of data analysis is the data cleaning where we will give data a perfect shape for the upcoming analysis.This step is very important before taking insights and data driven decision from the data.

First we will import python libraries which are used for data cleaning and dealing effectively with data 

```python
import pandas as pd
```

```python
import os
```

```python
import re
```

```python
os.listdir()
```

```python
naukri_data = pd.read_csv('Naukri Jobs Data.csv')
```

```python
naukri_data.head()
```

# First and foremost we will drop all the NaN values from the database 

```python
non_df = naukri_data[naukri_data.isna().any(axis = 1)]
non_df.head(50)

naukri_data = naukri_data.dropna(how='all')
naukri_data.head()
```

# Now we will separate the maximum years of experience required for certain job from certain company For this , we will use .str function to extract desired number of characters from the table

```python
naukri_data['max_exp_years'] = naukri_data['exp_required'].str[-6:-3]
naukri_data.head()
```

# As we can see after extracting the maximum years of exterience from the data we have got '-' sign additionalinto every two digit of experience years we extracted.To get rid off this sign we will use lambda and strip  function

```python
naukri_data['max_exp_years'] = naukri_data['max_exp_years'].map(lambda x: x.lstrip('-').rstrip('-'))
naukri_data.head(60)
```

# Now we will extract number of days from the Posted_as_on_22_5_2022 column of the table and then substract them from the date on which data is collected 

```python
new = naukri_data["Posted_as_on_22_5_2022"].str.split(" ", n = 1, expand = True)
naukri_data["day_of_posting"]= new[0]
naukri_data
```

# Then we will check and review our data randomly 

```python
naukri_data.iloc[15000]
```

```python
naukri_data.iloc[32737]
```

```python
naukri_data.iloc[23456]
```

```python
naukri_data.iloc[63]
```

```python
naukri_data['day_of_posting'] = pd.to_numeric(naukri_data['day_of_posting'])
```

# Yes !!! we got the problem , there was a string which was barring us from converting the day_of_posting column into numeric column that string was "TODAY" so we will convert that today to 0 number of days

```python
naukri_data['day_of_posting'] = naukri_data['day_of_posting'].replace(["TODAY"],"0")
naukri_data.head(24527)
```

```python
naukri_data[naukri_data['day_of_posting']== "TODAY" ].head(50)
```

```python
naukri_data['day_of_posting'] = naukri_data['day_of_posting'].str.strip()
naukri_data.head(253)
```

# Now we will remove all the data posted a month befor the given date is that 22 May 2022 so that we could properly take out insights from the data of a whole month 

```python
naukri_data.drop(naukri_data[naukri_data['day_of_posting'] == "30+"].index, inplace = True)
naukri_data.head(253)
```

```python
naukri_data['day_of_posting'] = naukri_data['day_of_posting'].replace(["TODAY"],"0")
naukri_data.head(24527)
```

# Now we will substract the number of days from the posted date to get exact date on which Job was posted on Naukri.com

```python
naukri_data['date_of_posting'] = pd.to_datetime('2022-05-22') - pd.to_timedelta(naukri_data['day_of_posting'], unit='d')
naukri_data
```

# Now we will remove the  "Day of Posting"  Column as we have nothing to do with it

```python
naukri_data.drop(['day_of_posting'], axis = 1, inplace = True)
```

```python
naukri_data
```

```python
naukri_data.drop(['Posted_as_on_22_5_2022'], axis = 1, inplace = True)
naukri_data
```

# Now we will extract minimum numbers of years of experience a candidate must have to apply for this Job

```python
naukri_data['min_exp_years'] = naukri_data['exp_required'].str[:2]
naukri_data
```


# Then cleaning it

```python
naukri_data['min_exp_years'] = naukri_data['min_exp_years'].map(lambda x: x.lstrip('-').rstrip('-'))
naukri_data
```

```python
naukri_data.drop(['exp_required'], axis = 1, inplace = True)
naukri_data
```

# As we know reviews make a huge weightage in any job in demand so we will extract number of reviews from the company_review by cleaning all non numeric characters from the column so that we could do furthur analysis on the data

```python
new2 = naukri_data["company_review"].str.split(" ", n = 1, expand = True)
naukri_data["companyreview"]= new2[0]
naukri_data
```

```python
naukri_data['companyreview3'] = naukri_data['companyreview'].map(lambda x: x.lstrip('('))

naukri_data
```

```python
naukri_data.drop(['company_review','companyreview'], axis = 1, inplace = True)
naukri_data
```

```python
naukri_data.rename(columns = {'companyreview3':'company_review'}, inplace = True)
naukri_data
```

```python
naukri_data['required_skills'] = naukri_data['required_skills'].replace(r'\n',' , ', regex=True) 
naukri_data
```
