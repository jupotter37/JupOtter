```python
#Since name was not present we need some way to identify different students so we are story their index in names
df.loc[df['Name'].isnull(), 'Name'] = df[df['Name'].isnull()].index.values
```

```python
additional_skills = set()

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Check if the value in the "Other skills" column is NaN
    if isinstance(row['Other skills'], str):
        # Split the values in the "Other skills" column by commas
        skills = row['Other skills'].split(',')
        # Add each skill to the set of additional skills
        additional_skills.update(skills)

# Print the extracted additional skills
print("Additional Skills:")
for skill in additional_skills:
    print(skill)
```

```python
machine_learning_skills = [
    'r programming',
    'r',
    'sql',
    'postgresql',
    'java',
    'c++ programming',
    'c++',
    'javascript',
    'pytorch',
    'image processing',
    'computer vision',
    'computervision',
    'big data analytics',
    'bigdata',
    'big data',
    'google cloud computing',
    'cloud computing',
    'git',
    'github',
    'linux',
    'aws (amazon web services)',
    'aws',
    'azure',
    'amazon web services',
    'microsoft azure',
    'azure (microsoft azure)',
    'hadoop',
    'spark',
    'flask',
    'django'
]
```

```python
for skill in machine_learning_skills:
    df[skill] = 0

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Check if the value in the "Other skills" column is a string
    if isinstance(row['Other skills'], str):
        # Split the values in the "Other skills" column by commas
        other_skills = row['Other skills'].split(',')
        # Update the corresponding skill column value to 1 if the student has that skill
        for skill in other_skills:
            skill = skill.strip().lower()  # Remove leading/trailing spaces and convert to lowercase
            if skill in machine_learning_skills:
                df.at[index, skill] = 1
```

```python
df['Are you available for 3 months, starting immediately, for a full-time work from home internship? '].unique()
```

```python
def calculate_availability_score(text):
    if "Yes" in text:
        return 1
    elif "No" in text:
        if "exams" in text or "college" or 'part' or 'available' or 'college' in text:
            return 0.5
        else:
            return 0
    else:
        return -1  # Use -1 for unknown or invalid values

# Create the availability score column
df['Availability Score'] = df['Are you available for 3 months, starting immediately, for a full-time work from home internship? '].apply(calculate_availability_score)
```

```python
# Define the weights for each column
column_weights = {
    'Python (out of 3)': 3,
    'Machine Learning (out of 3)': 3,
    'Natural Language Processing (NLP) (out of 3)': 3,
    'Deep Learning (out of 3)': 3,
    'r programming': 1,
    'r': 1,
    'sql': 1.5,
    'postgresql': 1.5,
    'java': 1,
    'c++ programming': 0.5,
    'c++': 0.5,
    'javascript': 1,
    'pytorch': 1,
    'image processing': 2,
    'computer vision': 2,
    'computervision':2,
    'big data analytics': 1,
    'bigdata': 1,
    'big data': 1,
    'google cloud computing': 1.5,
    'cloud computing': 1.5,
    'git': 1,
    'github': 1,
    'linux': 1,
    'aws (amazon web services)': 1.5,
    'aws': 1.5,
    'azure': 1.5,
    'amazon web services': 1.5,
    'microsoft azure': 1.5,
    'azure (microsoft azure)': 1.5,
    'hadoop': 1,
    'spark': 1.5,
    'flask': 1.5,
    'django': 1.5,
    'Availability Score': 3
}

# Calculate the overall score for each student
df['Overall Score'] = df.apply(lambda row: sum(row[column] * column_weights[column] for column in df.columns[1:-1]), axis=1)
df_copy['Overall Score'] = df['Overall Score']
# Sort the students based on their overall scores
df_sorted = df.sort_values('Overall Score', ascending=False)

# Reset the index of the sorted DataFrame
df_sorted.reset_index(drop=True, inplace=True)
```

```python
fig = px.scatter(df_sorted, x='Overall Score', y='Name', color='Overall Score',
                 hover_data=['Name', 'Overall Score'], labels={'Overall Score': 'Overall Score'},
                 title='Overall Score vs Name', color_continuous_scale='mrybm')

fig.update_layout(xaxis=dict(title='Overall Score'), yaxis=dict(title='Name'))

fig.show()
```
