```python
# Columns to convert to dummy variables
columns_to_convert = [
    'can work long time before system?', 'self-learning capability?',
    'Extra-courses did', 'talenttests taken?', 'olympiads',
    'Taken inputs from seniors or elders', 'interested in games',
    'In a Realtionship?', 'worked in teams ever?', 'Introvert', 'Job/Higher Studies?',
    'Gentle or Tuff behaviour?', 'Management or Technical', 'Salary/work', 'hard/smart worker'
]

# Convert columns to dummy variables
dummy_columns = pd.get_dummies(df[columns_to_convert])

# Concatenate the dummy columns with the original DataFrame
df = pd.concat([df, dummy_columns], axis=1)

# Drop the original columns (optional)
df.drop(columns=columns_to_convert, inplace=True)
```

```python
df
```

```python
# Convert object type columns to category type
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder

# Find columns with more than two unique categories
multi_category_columns = [col for col in df.columns if df[col].dtype == 'category' and len(df[col].cat.categories) > 2]

# Apply Label Encoding to columns with multiple unique categories
label_encoder = LabelEncoder()
for col in multi_category_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df['Suggested Job Role'].value_counts()
```

```python
X = df[['Acedamic percentage in Operating Systems', 'percentage in Algorithms', 'Percentage in Programming Concepts', 'Percentage in Software Engineering', 'Percentage in Computer Networks', 'Percentage in Electronics Subjects', 'Percentage in Computer Architecture', 'Percentage in Mathematics', 'Percentage in Communication skills', 'Type of company want to settle in?', 'Interested Type of Books', 'coding skills rating', 'Interested subjects', 'Logical quotient rating', 'Hours working per day', 'certifications', 'workshops']]
y = df['Suggested Job Role']
```

```python
# List all columns that contain the word 'Percentage' or 'percentage'
percentage_columns = [col for col in X.columns if 'percentage' in col.lower()]

# Divide each value in the percentage columns by 100
X[percentage_columns] = X[percentage_columns] / 100
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

# Apply SMOTE to generate synthetic samples
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X
```

# Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
# Splitting the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

prediction = decision_tree.predict(X_test)

print(classification_report(y_test, prediction))
print('\n')
print(confusion_matrix(y_test, prediction))
```

```python
# Predictions
y_pred_dt = decision_tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy score for Decision Tree
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Accuracy (Decision Tree): {accuracy_dt:.4f}")
```

## Random forest

```python
# Random Forest Classifier
model = RandomForestClassifier(n_estimators=500, random_state=42)
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)

print(classification_report(y_test, prediction))
print('\n')
print(confusion_matrix(y_test, prediction))
```

```python
import tensorflow as tf
from tensorflow import keras

class myCallback(tf.keras.callbacks.Callback):
        # Define the correct function signature for on_epoch_end
        def on_epoch_end(self, epoch, logs={}):
            if logs.get('accuracy') is not None and logs.get('accuracy') > 0.99:
                print("\nReached 99% accuracy so cancelling training!")

                # Stop training once the above condition is met
                self.model.stop_training = True

callbacks = myCallback()
```

```python
input_shape = X_train.shape[1]
print(input_shape)
```

```python
'''# Define the neural network architecture with L2 regularization
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(17,), kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dense(16, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dense(34, activation='softmax')  # Output layer with 34 units for classification
])'''
# Define the neural network architecture with two additional hidden layers
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(17,)),
    tf.keras.layers.Dense(32, activation='relu'),  # First additional hidden layer with 32 units
    tf.keras.layers.Dense(16, activation='relu'),  # Second additional hidden layer with 16 units
    tf.keras.layers.Dense(34, activation='softmax')  # Output layer with 34 units for classification
])
```

```python
# Increase the learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)

# Compile the model with the new optimizer
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#history = model.fit(X_train, y_train, epochs=5, validation_split=0.1, batch_size=32, callbacks=[callbacks])
# Train the model
model.fit(X_train, y_train, epochs=40, batch_size=50, validation_split=0.1,callbacks=[callbacks])
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```
