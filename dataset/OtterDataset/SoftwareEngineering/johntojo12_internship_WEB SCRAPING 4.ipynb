```python
#Making empty list to scrap the required details
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
#scrapping the rank of the video
try:
    ranks=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranks:  
        Rank.append(i.text)
        
except NoSuchElementException:
    Rank.append("-")
except StaleElementReferenceException:
    Rank.append("-")
print(Rank)
```

```python
#scrapping the name of the video

try:
    names=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        Name.append(i.text.split(sep='[')[0].replace('"',''))
        
except NoSuchElementException:
    Name.append("-")
except StaleElementReferenceException:
    Name.append("-")
print(Name)
```

```python
#scrapping the details of upload date

try:
    upload_dates=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_dates:
        Upload_date.append(i.text)
        
except NoSuchElementException:
    Upload_date.append("-")
except StaleElementReferenceException:
    Upload_date.append("-")
print(Upload_date) 
```

```python
#scrapping the details of views

try:
    view=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in view:
        Views.append(i.text)
        
except NoSuchElementException:
    Views.append("-")
except StaleElementReferenceException:
    Views.append("-")
print(Views)    
```

```python
#check
print(len(Rank),len(Name),len(Artist),len(Upload_date),len(Views))
```

```python
#creating dataframe
video_youtube=pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,"Upload Date":Upload_date,"Views":Views})
video_youtube
```

```python
for i in range(0,7):
    print(i)
    try:
        dt = driver.find_elements(By.XPATH,'//table[@border="1"]/tbody/tr/td[{}]'.format(i+1))
        for d in dt:
            datas[i].append(d.text)
            
    except NoSuchElementException:
        datas.append("-")
        
    except StaleElementReferenceException:
        datas.append("-")        

datas    
```

```python
#creating dataframe
# running from 1 because 0th index stores the name of the table 
dataset=pd.DataFrame({"Dataset name":datas[0][1:],"Data type":datas[1][1:],"Task":datas[2][1:],
                      "Attribute type":datas[3][1:],"No of instances":datas[4][1:],'No of attribute':datas[5][1:],
                      'Year':datas[6][1:]
                     })
dataset
```
