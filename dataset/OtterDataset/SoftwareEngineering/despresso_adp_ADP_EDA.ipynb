#### 데이터 추출(계통)

```python
[2*i for i in [1,2,3,4,5]]
```

```python
[2*i for i in range(10)]
```

```python
sel_index = [5*i for i in range(len(df)//5 + 1)]
df.iloc[sel_index]
```

```python
Q1 = df['Founded'].quantile(0.25)
Q3 = df['Founded'].quantile(0.75)
IQR = Q3 - Q1 
```

```python
df_after_IQR = df[(df['Founded'] <= Q3 + 1.5 * IQR) & (df['Founded'] >= Q1 - 1.5 * IQR)]
df_after_IQR.tail()
# df_after_IQR.shape
```
