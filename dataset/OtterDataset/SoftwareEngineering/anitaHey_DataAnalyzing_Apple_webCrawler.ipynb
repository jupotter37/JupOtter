```python
import selenium.webdriver.support.ui as ui
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import csv
import pandas
```

```python
def _extract(id_name: str):
    """ Extracts the specified element by class name """
    return driver.find_element_by_id(id_name).text
```

```python
def scrape_job(url: str, wait: ui.WebDriverWait):
    """ Scrape the job info from the specified Url. A broswer driver MUST be initialized beforehand.
    :param url: the url of a detailed google job page.
    :param wait: contains timeout.
    :return: a dict wrapping all info.
    """
    driver.get(url)
    
    # Wait until all required elements are generated.
    
    wait.until(lambda driver:EC.presence_of_element_located((By.ID, 'jdPostingTitle')))
    wait.until(lambda driver:EC.presence_of_element_located((By.ID, 'job-location-name')))
    wait.until(lambda driver:EC.presence_of_element_located((By.ID, 'accordion_keyqualifications')))
    wait.until(lambda driver:EC.presence_of_element_located((By.ID, 'accordion_additionalrequirements')))
    wait.until(lambda driver:EC.presence_of_element_located((By.ID, 'accordion_description')))
    wait.until(lambda driver:EC.presence_of_element_located((By.ID, 'accordion_education&experience')))
    
    try:
        sleep(2)
    # Extract job information.
        title = _extract("jdPostingTitle")
        location = _extract("job-location-name")
        minimum_qual = _extract("accordion_keyqualifications")
        
        try:
            preferred_qual = _extract("accordion_additionalrequirements")
        except:
            preferred_qual=""

        try:
            responsibilities = _extract("accordion_description")
        except:
            responsibilities=""
     
        try:
            experience = _extract("accordion_education&experience")
        except:
            experience=""
            
        return {
        'title': title,
        'location': location,
        'minimum_qual': minimum_qual,
        'preferred_qual': preferred_qual,
        'responsibilities': responsibilities,
        'education&experience':experience
        }
    except:
        return None
    
    
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.apple.com/jobs/us/');
sleep(1)
driver.find_element_by_id('page-search__input').send_keys('software')
driver.find_element_by_id('page-search__input').send_keys(Keys.RETURN)
driver.implicitly_wait(10) 
# print(driver.page_source)  # 顯示全部html
wait = ui.WebDriverWait(driver,20, 0.5)
arr = []

while True:
    try:
        sleep(2)
        wait.until(lambda driver: EC.presence_of_element_located((By.XPATH,"//a[@class='table--advanced-search__title']")))
        ans =  driver.find_elements_by_xpath("//a[@class='table--advanced-search__title']")
        for a in range(0,len(ans),1):
            arr.append(ans[a].get_attribute('href'))
        
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "pagination__next")))
        test = driver.find_element_by_class_name('pagination__next')
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "pagination__next")))
        next = test.find_element_by_tag_name('a')
        if next.get_attribute('aria-disabled') != 'true':
            next.send_keys(Keys.RETURN)
        else:break
    except:
        break
    
```

```python
len(arr)
```

```python
jobs = []
for url in arr:
    data = scrape_job(url, wait)
    if data !="":
        jobs.append(data)
        print(len(jobs))
driver.quit()
```

```python
csv_file = 'apple_jobs.csv'

with open(csv_file, 'w',newline='',encoding='utf-8') as f:
    w = csv.DictWriter(f, fieldnames = ['title', 'location', 'minimum_qual', 'preferred_qual', 'responsibilities','education&experience'])
    w.writeheader()
    
    for job in jobs:
        if job is not None:
            w.writerow(job)
    
print('File written: ' + csv_file)
```
