### Extract Image from Pdf

```python
# !pip install PyMuPDF # PyMuPDF: They will read the pDF images page by page
# !pip install pillow  # pillow is to extract the image
```

```python
import fitz
import PIL.Image # Python Image Library to Image
import io # io is string io or bytes, each bytes they can read it
```

```python
pdf = fitz.open("C:\\Users\\hp\\Desktop\\Projects\\Machine Learning Model Deployment.pdf")
counter = 1

for i in range(len(pdf)):
    page = pdf[i]
    images = page.get_images()
    for image in images:
        base_img = pdf.extract_image(image[0])
        image_data = base_img["image"]
        img = PIL.Image.open(io.BytesIO(image_data))
        extension = base_img['ext']
        img.save(open(f'image{counter}.{extension}', "wb"))
        counter +=1
```
