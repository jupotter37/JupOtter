** 7) List Comprehension **
* **One of the most important topic of this kernel .**
* We use list comprehension for data analysis often. 
* list comprehension: collapse for loops for building lists into a single line 
* Ex: num1 = [2,3,4] and we want to make it num2 = [3,4,5]. This can be done with for loop. However it is unnecessarily long. We can make it one line code that is list comprehension.

```python
# Example of list comprehension
num1 = [2,3,4]
num2 = [i + 1 for i in num1 ]
print(num2)
```

* [i + 1 for i in num1 ]: list of comprehension 
* i +1: list comprehension syntax 
* for i in num1: for loop syntax 
* i: iterator 
* num1: iterable object

```python
# lets return world happiness report csv and make one more list comprehension example
threshold = sum(data["Happiness Score"])/len(data["Happiness Score"])
data["score_level"] = ["high" if i > threshold else "low" for i in data["Happiness Score"]]
data.loc[:10,["score_level","Happiness Score"]] 
```

**B. Scope : **  The scope of a variable refers to the places that you can see or access a variable.

* global: defined main body in script
* local: defined in a function
* built in scope: names in predefined built in scope module such as print, len 

```python
x = 5
def f():
    x = 7
    return x
print(x)      # x = 5 global scope
print(f())    # x = 7 local scope
```

```python
# What if there is no local scope
x = 5
def f():
    y = 10+x        # there is no local scope x
    return y
print(f())         # it uses global scope x
# First local scopesearched, then global scope searched, if two of them cannot be found lastly built in scope searched.
```

**C. Nested Function**
* A function defined inside another function is called a nested function. 

```python
def number(x):
    def add(y):
        return x + y
    return add

output = number(5)
print(output(12)) 
```

**D.  Default and flexible arguments**
*  Default Arguments: Function arguments can have default values in Python. We can provide a default value to an argument by using the assignment operator (=).  Default argument example: 
def f(a, b=1):
>   """ b = 1 is default argument"""
*  Flexible arguments:  The special syntax, *args  in function definition is used to pass a variable number of arguments to a function. 
def f(*args):
>   """ *args can be one or more"""



```python
#Default Arguments
def x (a, b = 10):
    y = a*b 
    return y
print(x(5))
```

```python
#Flexible Arguments
# *args
def f(*args):
    z = 1
    for i in args:
        z *= i
    print(z)

f(4, 5)
f(2, 3, 4)
# We used *args to send a variable-length argument list to our function, we were able to pass in as many arguments as we wished into the function calls.
```

**E. LAMBDA FUNCTION**
* Faster way of writing function

```python
def calculate(x):
    output=x*x
    return output
print(calculate(3))
```

Instead of writing the above code, we can write faster using the lambda function.

```python
calculate = lambda x: x * x
# Output: 10
print(calculate(3))
```

**F. ANONYMOUS FUNCTİON**
*  Like lambda function but it can take more than one arguments.

   map(func,seq) : applies a function to all the items in a list

```python
number_list = [5,2,3]
y = map(lambda x:x**2,number_list)
print(list(y))
```
