## Q1. Scrape the details of most viewed videos on YouTube from Wikipedia


```python
#import all required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
import warnings
warnings.filterwarnings("ignore")

from selenium.common.exceptions import NoSuchElementException
```

```python
#connecting to webdriver
driver = webdriver.Chrome('chromedriver')
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
rank = []
name = []
artist = []
date = []
views =[]
```

```python
#let's extract all web elements having rank as parent tag
rank_tags = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"]')

#extracting rank
rank_test = []

for i in rank_tags:
    rank_test.append(i.text.split("\n"))
rank_try = rank_test[0][2:32]

rank = [j.split(".")[0] for j in rank_try]

len(rank)
```

```python
#let's extract all web elements having name as parent tag
name_tags = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"]')

#extracting name
name_test = []

for i in name_tags:
    name_test.append(i.text.split("\n"))
name_try = name_test[0][2:32]

name = [j.replace('"','.').split(".")[2] for j in name_try]

len(name)
```

```python
#let's extract all web elements having artist as parent tag
artist_tags = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"]')

#extracting artist
artist_test = []

for i in artist_tags:
    artist_test.append(i.text.split("\n"))
artist_try = name_test[0][2:32]

artist_try1 = [j.replace(']','.').split(".")[2] for j in artist_try]

artist = [j.rsplit(" ",1)[0] for j in artist_try1]

len(artist)
```

```python
#let's extract all web elements having date as parent tag
date_tags = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"]')

#extracting date
date_test = []

for i in date_tags:
    date_test.append(i.text.split("\n"))
date_try = date_test[0][2:32]

date_try1 = [j.replace('[','.').split(".")[3] for j in date_try]

date = [j.split(" ",1)[1] for j in date_try1]

len(date)
```

```python
#let's extract all web elements having views as parent tag
views_tags = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"]')

#extracting views
views_test = []

for i in views_tags:
    views_test.append(i.text.split("\n"))
views_try = views_test[0][2:32]

views_try1 = [j.split("[")[1].strip() for j in views_try]
views_try1

views = [j.rsplit(" ",4)[1] for j in views_try1]
len(views)
```

```python
youtube = pd.DataFrame()
youtube['Rank'] = rank
youtube['Name'] = name
youtube['Artist'] = artist
youtube['Upload Date'] = date
youtube['Views (in Billions)'] = views
youtube
```

```python
driver.close()
```

### Q5. Scrape the details of trending repositories on Github.com

```python
#connecting to webdriver
driver = webdriver.Chrome('chromedriver')
```

```python
url = 'https://github.com/'
driver.get(url)
```

```python
#Finding web element for explore tab
search_exp = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
search_exp.click()

#Finding web element for trending option in drop down
search_tr = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
search_tr.click()
```

```python
title = []
des = []
contri = []
language = []
```

```python
#let's extract all web elements having title as parent tag
title_tags = driver.find_elements(By.XPATH, '//h1[@class="h3 lh-condensed"]')

#extracting title
title = []
for i in title_tags:
    title.append(i.text)
    
len(title)
```

```python
#Fetching url to open each description page
url = []
url_tags= driver.find_elements(By.XPATH, '//h1[@class="h3 lh-condensed"]/a')
for i in url_tags:
    url.append(i.get_attribute('href'))
len(url)
```

```python
import time

#let's extract all web elements having required parent tags and then extract required elements from them
for i in url:
    driver.get(i)
    time.sleep(3)
    
    #Extracting Description from parent tag
    try:
        des_tag = driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        des.append(des_tag.text)
    except NoSuchElementException:
        des.append("-")
    
    #Extracting Expected Contributors count from parent tag
    try:
        contri_tag = driver.find_elements(By.XPATH,'//h2[@class="h4 mb-3"]')
        contri_try=[]
        for j in contri_tag:
            contri_try.append(j.text)

        #to get string with substring(only 'Contributors') in it
        contri_try1 = [i for i in contri_try if "Contributors" in i]
        contri.append(contri_try1[0].split(" ")[1])
    except NoSuchElementException:
        contri.append("-")
    
    #Extracting Language used from parent tag
    try:
        language_tag = driver.find_elements(By.XPATH,'//li[@class="d-inline"]')
        language_try = []
        for k in language_tag:
            language_try.append(k.text.replace("\n"," : "))
        language.append(language_try)
        time.sleep(3)
    except NoSuchElementException:
        language.append("-")
        time.sleep(3)
```

```python
len(des)
```

```python
len(contri)
```

```python
len(language)
```

```python
git = pd.DataFrame()
git['Repository Title'] = title
git['Repository Description'] = des
git['Contributor count'] = contri
git['Language Used'] = language
git
```

```python
driver.close()
```

### Q10. Details of Datasets from UCI machine learning repositories

```python
#connecting to webdriver
driver = webdriver.Chrome('chromedriver')
```

```python
url = 'https://archive.ics.uci.edu/'
driver.get(url)
```

```python
#Finding web element for view all datasets option
search_data = driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]")
search_data.click()
```

```python
dname=[]
```

```python
#let's extract all web elements having Dataset Name as parent tag
dname_tags = driver.find_elements(By.XPATH, '//b/a')

#extracting Dataset Name
for i in dname_tags:
    dname.append(i.text)
len(dname)
```

```python
#Fetching url to open each description page
url = []
url_tags= driver.find_elements(By.XPATH, '//b/a')
for i in url_tags:
    url.append(i.get_attribute('href'))
len(url)
```

```python
dtype=[]
task=[]
atype=[]
inst=[]
attr=[]
year=[]
```

```python
import time

#let's extract all web elements having required parent tags and then extract required elements from them
for i in url:
    driver.get(i)
    time.sleep(2)
    dtype_test = []
    task_test = []
    atype_test = []
    inst_test=[]
    attr_test=[]
    year_test=[]
    
    #Extracting data type
    try:
        dtype_tag = driver.find_elements(By.XPATH,'//p[@class = "normal"]')
        for i in dtype_tag:
            dtype_test.append(i.text)
        dtype.append(dtype_test[2])
    except NoSuchElementException:
        dtype.append("N/A")
    
    #Extracting task
    try:
        task_tag = driver.find_elements(By.XPATH,'//p[@class = "normal"]')
        for i in task_tag:
            task_test.append(i.text)
        task.append(task_test[14])
    except NoSuchElementException:
        task.append("N/A")
        
    #Extracting Attribute type    
    try:
        atype_tag = driver.find_elements(By.XPATH,'//p[@class = "normal"]')
        for i in atype_tag:
            atype_test.append(i.text)
        atype.append(atype_test[8])
    except NoSuchElementException:
        atype.append("N/A")
    
    #Extracting No. of instances
    try:
        inst_tag = driver.find_elements(By.XPATH,'//p[@class = "normal"]')
        for i in inst_tag:
            inst_test.append(i.text)
        inst.append(inst_test[4])
    except NoSuchElementException:
        inst.append("N/A")
    
    #Extracting No.of attributes
    try:
        attr_tag = driver.find_elements(By.XPATH,'//p[@class = "normal"]')
        for i in attr_tag:
            attr_test.append(i.text)
        attr.append(attr_test[10])
    except NoSuchElementException:
        attr.append("N/A")
    
    #Extracting Year
    try:
        year_tag = driver.find_elements(By.XPATH,'//p[@class = "normal"]')
        for i in year_tag:
            year_test.append(i.text)
        year.append(year_test[12].split("-")[0])
    except NoSuchElementException:
        year.append("N/A")
        
    time.sleep(2)
```

```python
dset = pd.DataFrame()
dset['Dataset Name'] = dname
dset['Data type'] = dtype
dset['TAsk'] = task
dset['Attribute Type'] = atype
dset['No. of Instances'] = inst
dset['No. of Attribute'] = attr
dset['Year'] = year
dset
```

```python
driver.close()
```
