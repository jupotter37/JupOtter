```python
# Distribution of students by gender, age, and university
Gender_Disribution = df['gender'].value_counts()
Age_Distirbution = df['age'].value_counts()
University_Distribution = df['university'].value_counts()

print('Gender_Disribution:', Gender_Disribution)
print('Age_Distirbution:', Age_Distirbution)
print('University_Distribution:', University_Distribution)
```

```python
#universities with highest average of mental health indicators 
ranked_universities_mental_health = df.groupby('university')[['depression', 'anxiety', 'isolation']].mean().sort_values(by=['depression', 'anxiety', 'isolation'], ascending=False)
(ranked_universities_mental_health)
```

```python
#correlation between CGPA and mental health indicators
#since cgpa is a string and not numeric, we find midpoint of cgpa
def cgpa_to_midpoint(cgpa):
    low,  high = map(float, cgpa.split('-'))
    result = (low + high) / 2
    return result

df['cgpa_midpoint'] = df['cgpa'].apply(cgpa_to_midpoint)
```

```python
#now we calculate the correlation of cgpa to mental health factors
correlation_cgpa_mentalhealth = df[['cgpa_midpoint', 'depression', 'anxiety', 'isolation']].corr().loc[['cgpa_midpoint'], ['depression', 'anxiety', 'isolation']]
correlation_cgpa_mentalhealth
```

```python
#Impact of academic workload and pressure on mental health
academicworkload_impact = df.groupby(['academic_workload'])[['depression', 'anxiety', 'isolation']].mean()
academicpressure_impact = df.groupby(['academic_pressure'])[['depression', 'anxiety', 'isolation']].mean()

display(academicpressure_impact, academicworkload_impact)

```

```python
#Analysis of popular stress relief activities among students
df['stress_relief_activities_list'] = (
    df['stress_relief_activities']
    .str.split(', ')
)
encoded = (
    df['stress_relief_activities_list']
    .str.join('|')
    .str.get_dummies()
)
frequency = (
    encoded
    .sum()
    .sort_values(ascending=True)
)
print('Anti_Stress_Activities =\n', frequency)
```

```python
#Comparison of mental health metrics between undergraduate and postgraduate students
undergrad_postgrad_comparison = df.groupby('degree_level')[['depression', 'anxiety', 'isolation', 'future_insecurity']].mean()
undergrad_postgrad_comparison
```

```python
#Differences in mental health based on degree majors
DegreeMajor_comparison = df.groupby('degree_major')[['depression', 'anxiety', 'isolation', 'future_insecurity']].mean().sort_values(['depression', 'anxiety', 'isolation', 'future_insecurity'], ascending=False)
DegreeMajor_comparison
```
