```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('glassdoor_jobs.csv')
df
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
df.columns
```

```python
df['hourly']=df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df[['min_salary', 'max_salary']] = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K').astype(float).fillna(np.nan)


df.drop(columns=['Salary Estimate'], inplace=True)
```

```python
df
```

```python
df['min_salary'] =df['min_salary'].fillna(np.nan) * 1000
df['max_salary'] =df['max_salary'].fillna(np.nan) * 1000
df
```

```python
df['Company Name']=df.apply(lambda x:x['Company Name'] if x['Rating']<0 else x['Company Name'][:-3],axis=1)
```

```python
df['Job_state']=df['Location'].apply(lambda x:x.split(',')[1])
```

```python
df['avg_salary']=(df['min_salary']+
df['max_salary'])//2
```

```python
df['same_state']=df.apply(lambda x:1 if x.Location==x.Headquarters else 0,axis=1)
```

```python
df['age']=df.Founded.apply(lambda x:x if x<0 else 2024-x)
```

```python
#python
df['python']=df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['Rstudio'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
 
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)

df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)

df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df.tableau.value_counts()
```

```python
df 
```

```python
df['Sector'].unique()
df = df[df['Sector']!= '-1']
```

```python
#df=df.drop(['Job Description'],axis=1)
df=df.drop(['Location'],axis=1)
```

```python

df.drop(columns=['Salary Estimate'], inplace=True)
```

```python
df['Size'] = df['Size'].str.extract(r'(\d+)(?=\s+to|\+|\-| employees)').astype(float)
```
