# Import librairies

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

## Import data

```python
data=pd.read_csv("../data/cleaned_dataset.csv")
data=data.drop("Unnamed: 0", axis=1)
train_data, test_data = train_test_split(data, test_size=0.2)
```

## 1) Multiple linear regression using OLS

```python
X = train_data.drop("basesalary", axis=1)
Y = train_data["basesalary"]

regression = sm.OLS(Y, X).fit()
regression.summary()
```

```python
predictions = regression.predict(test_data.drop("basesalary", axis=1))
```

```python
rmse = float(format(np.sqrt(mean_squared_error(test_data["basesalary"], predictions)), '.3f'))
print("\nRMSE: ", rmse)
```

## 2) Random forest regression

```python
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0, max_depth=10)
regressor.fit(X, Y) 
y_pred = regressor.predict(X)
rmse = float(format(np.sqrt(mean_squared_error(Y, y_pred)), '.3f'))
print("\nRMSE: ", rmse)
```
