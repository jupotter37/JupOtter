```python
def salva_linhas_em_lista(nome_arquivo = "referencias ATDAnalyzer.txt"):
  lista_referencias = []
  with open(nome_arquivo, mode="r") as f_temp:
    for linha in f_temp:
      lista_referencias.append(linha.strip())
  return lista_referencias

# cria um novo arquivo com as referências
def cria_novo_arquivo(lista):
  novo_arquivo = "minhas_referencias.txt"
  with open(novo_arquivo, mode='w') as f_temp:
    for each in lista:
      conteudo = each + '\n'
      f_temp.write(conteudo)
```

```python
lista_temp = []
for each in lista_ok:
  conteudo = each.split("). ")
  temp = conteudo[1].split(". ")
  elemento = conteudo[0], temp[0]
  lista_temp.append(elemento)
```
