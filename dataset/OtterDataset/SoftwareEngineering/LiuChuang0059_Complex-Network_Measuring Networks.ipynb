# 聚类系数

> 在你的朋友关系网络中， 你的两个朋友也可能是朋友， 这种可能性的大小反映了朋友圈的紧密程度。 聚类系数： 定量刻画任意两个朋友之间也相互为朋友的概率。 

* 聚类系数： 包含中心节点的三角闭合数目和 以中心节点为中心的连通三元组的数目的比

* 平均聚类系数： 所有节点聚类系数的平均值

* 整个网络的聚类系数 ：网络中所有的三角闭合和可能存在的三角闭合之间的比值 

```python
cc = nx.clustering(nx.Graph(G), ego) # 中心网络的聚类系数
print(cc)
```

```python
acc = nx.average_clustering(nx.Graph(G)) # 平均网络聚类系数
print(acc)
```

```python
trans = nx.transitivity(G) # 整个网络的聚类系数  ： 横截性
print(trans)
```

> 上述考虑的都是无权网络，对于加权网络， 分析较为复杂

# Node eccentricity

* 一个节点到其他所有节点的距离

```python
ecc = nx.eccentricity(nx.Graph(G))
```

```python
print(ecc[ego])
```

**网络的直径就是最大的 eccentricity**


```python
print(nx.diameter(G,ecc))
```

* 网络的半径是最小偏心量。这个定义并不直观，但它就是这样。更违反直觉的是，一般来说，半径不是直径的一半。

```python
print(nx.radius(G,ecc))
```

**网络的中心是 离心距等于半径的节点的集合**

```python
nx.center(G,ecc)
```

**网络的外围是所有离心距等于直径的节点的集合。复杂网络中的外围节点集通常很大。**

```python
len(nx.periphery(G,ecc))
```

# 选择正确的中心

* 选出最重要的节点

* 一个节点的重要性取决于该节点在网络中的结构位置，可以通过邻近、最短路径或其他一些结构元素来计算。

##  度中心性

* 一个节点的度越大就意味着这个节点越重要

* 为了便于不同网络之间比较， 对中心性指标进行归一化处理 （不然大网络中 小度节点比小网络中大度节点还多）

$$DC= \frac{k_{i}}{N-1}$$



## 接近中心性

* 一个节点到网络中所有及诶单的距离的平均值 的倒数

* 值越大说明节点越接近其他的节点

* 接近数大的节点对于网络中信息的流动有着最佳的观察视野

## 介数中心性

* 经过某个节点的最短路径的数目

* 刻画了节点对网络中信息传输的控制能力



## 特征向量中心性

* 节点的重要性既取决于邻居节点的数目，也取决于邻居节点的重要性

* 基于这种想法有两种重要的算法： HITS 和 PageRank

# 中心性指标分析

## 中心性指标计算

```python
dgr = nx.degree_centrality(G)
clo = nx.closeness_centrality(G)
har = nx.harmonic_centrality(G)
eig = nx.eigenvector_centrality(G)
bet = nx.betweenness_centrality(G)
pgr = nx.pagerank(G)
hits = nx.hits(G)

```

```python
import pandas as pd
```

```python
centralities = pd.concat(
    [pd.Series(c) for c in (hits[1], eig, pgr, har, clo, hits[0], dgr, bet)],
    axis=1)

centralities.columns = ("Authorities", "Eigenvector", "PageRank",
                        "Harmonic Closeness", "Closeness", "Hubs",
                        "Degree", "Betweenness")
centralities["Harmonic Closeness"] /= centralities.shape[0]
```

```python
centralities
```

## 中心性指标之间关联

```python
c_df = centralities.corr()
ll_triangle = np.tri(c_df.shape[0], k=-1)
c_df *= ll_triangle
c_series = c_df.stack().sort_values()
c_series
```

# 通过协调性评估网络的均匀性

* 进行节点属性分析， 关注于协调性。

* 协调性指的是 连边的两端的两个节点的关联性是正向的还是负向的

* 正向的属性关联称为协调网络， 协调网络倾向于和有相同属性的节点呢连接

最简单的协调性是 节点的度（出度和入度）： 节点的度和她他的邻居的度之间的相互关系。

```python
my_degree, their_degree = zip(*nx.average_degree_connectivity(G).items())
```

> nx.average_degree_connectivity 返回一个字典： key是 自身的度， value 是邻居节点的度的平均值。

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.scatter( my_degree, their_degree)
```

* 如果网络是均匀的， 分布应该是 $y = x$

* 可以看到对于度大的节点， 不太均匀
