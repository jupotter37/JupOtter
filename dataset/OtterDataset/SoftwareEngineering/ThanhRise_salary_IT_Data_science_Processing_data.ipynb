```python
# Read in the data from multiple files
def read_data():
    data = pd.DataFrame()
    for file in os.listdir("data"):
        if file.endswith(".csv"):
            data = pd.concat([data, pd.read_csv("data/" + file)])
    return data
```

```python
# concat data_df and indee_df
data_df = pd.concat([data_df, indee_df], ignore_index=True)
data_df
```

```python
# remove duplicates and reset index
data_df = data_df.drop_duplicates().reset_index(drop=True)
```

```python
# fill -1 with NaN
data_df = data_df.replace(-1, np.nan)
data_df = data_df.replace('-1', np.nan)
data_df = data_df.replace('None', np.nan)
data_df = data_df.replace( -1.0, np.nan)
data_df = data_df.replace( '--', np.nan)
data_df = data_df.replace( '', np.nan)
```

```python
data_df['Location'].unique()
```

```python
# split Location into City and State if there is a comma else set city to NaN
data_df = data_df.join(data_df['Location'].str.split(',', 1, expand=True).rename(columns={0:'City', 1:'State'}))
```

```python
# change State to remote if city is remote
data_df.loc[data_df['City'] == 'Remote', 'State'] = 'Remote'
data_df['State'].unique()
```

```python
# drop salary estimate nan
data_df = data_df.dropna(subset=['Salary Estimate'])
```

```python
# remove (Employer est.) from salary estimate
data_df['Salary Estimate'] = data_df['Salary Estimate'].str.replace('(Employer est.)', '')
data_df['Salary Estimate'] = data_df['Salary Estimate'].str.replace('(Glassdoor est.)', '')
```

```python
# remove '()' from salary estimate
data_df['Salary Estimate'] = data_df['Salary Estimate'].str.replace('(', '')
data_df['Salary Estimate'] = data_df['Salary Estimate'].str.replace(')', '')
```

```python
data_df
```

```python
def salary_base(x):
    y=""
    if  "month" in str(x) or "Month" in str(x):
        y = "Monthly"
    elif "hour" in str(x) or "Hour" in str(x):
        y = "Hourly"
    elif "day" in str(x) or "Day" in str(x) or "per day" in str(x) or "Daily" in str(x) or "daily" in str(x):
        y = "Daily"
    else:
        y = "Yearly"
    
    return y 
        
```

```python
data_df['Salary Base'] = data_df['Salary Estimate'].apply(salary_base)
```

```python
# split salary estimate into min and max, if there is a - else set max and min to NaN
data_df = data_df.join(data_df['Salary Estimate'].str.split('-', 1, expand=True).rename(columns={0:'Salary Min', 1:'Salary Max'}))
data_df['Salary Min'] = data_df['Salary Min'].str.replace('$', '')
data_df['Salary Max'] = data_df['Salary Max'].str.replace('$', '')
```

```python
# change K to ,000 in Salary Min and Salary Max
data_df['Salary Min'] = data_df['Salary Min'].str.replace('K', '000')
data_df['Salary Max'] = data_df['Salary Max'].str.replace('K', '000')
# if Max is None then set Max to Min and Min to NaN
data_df.loc[data_df['Salary Max'] == 'None', 'Salary Max'] = data_df['Salary Min']
```

```python
def to_float(x):
    if x is not None:
        x = str(x)
        x = re.sub('[^0-9.,]', '', x)
        x = x.replace(',', '')
        x = x.replace(' ', '')
        x = float(x)
    else:
        x = np.nan
    return x
```

```python
data_df['Salary Min'] = data_df['Salary Min'].apply(to_float)
data_df['Salary Max'] = data_df['Salary Max'].apply(to_float)
```

```python
# if max is nan then set max to min
data_df.loc[data_df['Salary Max'].isnull(), 'Salary Max'] = data_df['Salary Min']
```

```python
# returned to the same scale

def adjust_salary2(m,n):
    salary = []
    for x, y in zip(m,n):
        if x == "Yearly":
            yearly_salary = y
        if x == "Monthly":
            yearly_salary = y*12
        if x == "Hourly":
            yearly_salary = y*40*52
        if x == "Daily":
            yearly_salary = y*5*52
            
        salary.append(yearly_salary)
        
    return salary
```

```python
data_df['Salary Min'] = adjust_salary2(data_df['Salary Base'], data_df['Salary Min'])
data_df['Salary Max'] = adjust_salary2(data_df['Salary Base'], data_df['Salary Max'])
```

```python
# process job title and job description
def clen_text(text):
    # lowercase text
    text = text.lower()
    # remove punctuation to ' '
    text = text.translate(str.maketrans('', '', string.punctuation))
    # remove words containing numbers
    text = re.sub('\w*\d\w*', ' ', text)
    # remove new line characters
    text = re.sub('\n', ' ', text)
    # remove non-ascii characters
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # remove single characters
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
data_df['Job Title'] = data_df['Job Title'].apply(clen_text)
data_df['Job Description'] = data_df['Job Description'].apply(clen_text)
```
