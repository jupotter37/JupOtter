```python
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException  # Importing NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
```

```python
url1 = 'https://www.linkedin.com/my-items/saved-jobs/?cardType=APPLIED'
```

```python

# Create a Service object
service = Service()

# Initialize the WebDriver with the Service object
driver = webdriver.Chrome(service=service)
driver.implicitly_wait(10)


driver.get(url1)
```

```python
def scroll_down():
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Add a short delay to allow content to load
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException

import time  # Import the time module for adding delays

# Main loop to scroll through and extract job data
while True:
    # Scroll down to load more job listings
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    print("Scrolled down")
    time.sleep(2)  # Add a delay to allow time for the new job listings to load

    # Find the list containing the job listings
    job_list = driver.find_element(By.CLASS_NAME, 'reusable-search__entity-result-list')
    print("Found job list")

    # Find all job elements within the list
    job_elements = job_list.find_elements(By.CLASS_NAME, 'reusable-search__result-container')
    print(f"Found {len(job_elements)} job elements")

    # Extract data from each job element
    for job in job_elements:
        position = job.find_element(By.CLASS_NAME, 'entity-result__title-text').text
        company = job.find_element(By.CLASS_NAME, 'entity-result__primary-subtitle').text
        location = job.find_element(By.CLASS_NAME, 'entity-result__secondary-subtitle').text
        jobs_data.append({
            'Position': position,
            'Company Name': company,
            'Location': location
        })

    # Look for the next button
    next_button = driver.find_element(By.CLASS_NAME, 'artdeco-pagination__button--next')
    print("Found next button")
    
    # Check if the next button is disabled
    if 'disabled' in next_button.get_attribute('class'):
        print("Next button disabled. Exiting the loop.")
        break  # Exit the loop if the "Next" button is disabled (reached the last page)

    # Click the "Next" button to navigate to the next page
    actions = ActionChains(driver)
    actions.move_to_element(next_button).perform()  # Scroll to the "Next" button
    next_button.click()  # Click the "Next" button
    print("Clicked next button")
    time.sleep(2)  # Add a delay after clicking the "Next" button

# Print the scraped job data
for job in jobs_data:
    print(job)

# Close the WebDriver
driver.quit()
```

```python
df = pd.DataFrame(jobs_data,columns = ['Position',"Company Name","Location"])
```
