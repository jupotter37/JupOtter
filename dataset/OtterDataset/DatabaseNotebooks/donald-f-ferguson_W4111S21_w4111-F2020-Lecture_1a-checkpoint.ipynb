__What is the Maximum Number of Predicted Deaths for the United States of America?__

````python
q_result = %sql select max(totdea_mean) as oh_darn from ihmefast where location_name = "United States of America"
oh_darn = q_result[0]["oh_darn"]
print("\n\nThe worst prediction for the US total deaths is", round(oh_darn,0), "\n")
````

- SQL statement: ```select max(totdea_mean) as oh_darn from ihmefast where location_name = "United States of America"```
    
    
- What does that statement mean?
    - Scan all rows in the database table.
    - For all rows with column named ```location_name``` with value "United States of America"
    - Find the maximum value for the columna ```totdea_mean.```
    

__Which Location has the Worst Predicted Deaths?__


````python
%sql select location_name, round(max(totdea_mean), 0) as total_deaths from ihmefast group by location_name order by total_deaths desc limit 10;
````

__How has Maximum Death Prediction for USA Changed over Time?__


````python
%%sql

select belief_date, round(max(totdea_mean), 0) as total_deaths from ihmefast
    where location_name="United States of America"
    group by belief_date order by belief_date;
````

__What Locations do I Know About?__

````python
import time
start_time = time.time()

# This is the actual statement. The rest is just timing it.
res = %sql select distinct location_name from ihmefast


elapsed_time = time.time() - start_time
print("Distinct locations are: \n", res)
print("Total elapsed time = ", round(elapsed_time, 3), "seconds")
````

- SQL statement: ```select distinct location_name from ihmefast```


- Is equivalent to the following code, which I ran against a CSV with all of the records.


````python
import csv

def get_distinct_column_values(f_name, col_name, trace_count=0):

    result = set()
    rows_read = 0

    with open(f_name, "r") as in_file:
        
        result = set()

        csv_file = csv.DictReader(in_file)

        for r in csv_file:
            rows_read += 1
            if trace_count > 0 and rows_read % trace_count == 0:
                print("Have read ", rows_read, "rows.")

            val = r.get(col_name, None)
            if val is not None:
                result.add(val)
                
    return result
````

- Let's do a timed run of the code that I wrote against a single CSV file that contains all of the entries.

````python
d_file = "/Users/donaldferguson/Dropbox/Columbia/Courses/W4111F2020/IntroductionToDatabases_F20/Data/ihmefast.csv"
start_time = time.time()
res = get_distinct_column_values(d_file, "location_name", 0)
elapsed_time = time.time() - start_time
print("Distinct locations are: \n")
for r in res:
    print(r)
print("\n\notal elapsed time = ", round(elapsed_time, 3), "seconds")
````

- Some observations:
    - The SQL statement was a lot more compact and simple.
    - The execution time was about 4,000 times faster.

    - Specialize "query languages."
    - Performance optimization for data centric applications.

- What are the cities in Massachusetts? 

````python
%%sql ma_towns <<
        select * from 
            (select 
                countries.name as country_name, countries.iso2 as country_code, 
                cities.name as city_name, 
                (select states.name as state_name from states where 
                     states.id=cities.state_id) as state_name 
                from 
                    countries join cities on countries.id = cities.country_id) as b 
        where state_name='Massachusetts';
````
