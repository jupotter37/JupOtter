```python
job_dec=[]
```

```python
import re

def replace_multiple(text, replacements):
    # Create a regular expression from the keys of the replacements dictionary
    regex = re.compile("|".join(map(re.escape, replacements.keys())))
    
    # Replace function using the dictionary
    return regex.sub(lambda match: replacements[match.group(0)], text)
```

```python
para = '''Recruitment and sourcing strategies
Candidate assessment and screening
Interviewing techniques
Employment law and regulations
Negotiation and offer management
Candidate relationship management (CRM)
ATS (Applicant Tracking System) proficiency (e.g., Greenhouse, Lever, Workday)
Networking and relationship building
Employer branding and marketing
Data-driven decision making
HRIS (Human Resource Information System) familiarity'''




# eleminat the fault patterns from the paragraph of job description
rep = {   
    ": ":",",
    " (":",",
    ") ":",",
    ")":",",
    "\n":",",
    ",  ":",",
    ", ":",",
    ",,":",",
    ", ,":",",
    " and ":",",
    "  ":"" ,
    "for ":""
    }
```

```python
a = replace_multiple(para, rep)
```

```python
# split the skills into list from paragraph
b=a.split(',')
```

```python
b
```

```python
# append list of skills of specific job roles
job_dec.append(b)
job_dec
```

```python
# pickle of job description
import pickle

# Specify the file name
file_name = 'job_dec.pkl'

# Write list to pickle file
with open(file_name, 'wb') as file:
    pickle.dump(job_dec, file)

print(f"List has been written to '{file_name}' using pickle.")
```
