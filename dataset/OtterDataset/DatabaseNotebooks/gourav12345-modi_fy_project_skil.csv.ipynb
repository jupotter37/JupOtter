```python
import pandas as pd
```

```python
data= pd.read_csv("/home/gourav/MyData/cs/fy_project/skills.csv")
data
```

```python
data.columns
```

```python
seColumnList = data['Sofware Engineer'].tolist()
seColumnList
```

```python
skills = []
for skillStr in seColumnList:
   
    if (pd.isna(skillStr)):
        continue
    val = skillStr[2:-2]
    val = val.split(', ')
    for v in val:
        skillArr = v.split("\\n")
        for skill in skillArr:
            sk = skill.split(':')
            
            for s in sk:
                if s!='' and len(s) <40 and s not in skills:
                    skills.append(s)
                    print(s)
```

```python
len(skills)
```

```python
df = pd.DataFrame(columns=skills)
df
```

```python
for user in seColumnList:
    userSkills = []
    for skill in skills:
        if isinstance(user, str) and skill in user:
            userSkills.append(1)
        else:
            userSkills.append(0)
    df.loc[len(df.index)]= userSkills
```

```python
df
```

```python
def jaccard(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```

```python
jaccard(df.iloc[0].tolist(), df.iloc[1].tolist())
```
