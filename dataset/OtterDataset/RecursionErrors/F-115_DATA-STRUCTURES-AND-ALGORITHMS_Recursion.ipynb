```python
# Task 1(a)
# Factorial of num

def fact(num):

    if num == 0 or num == 1:  # Base condition

        return 1

    else:

        return num * fact(num - 1)


f1 = fact(6)
print(f1)







# Task 1(b)
# n'th Fibonacci number.


def fibonacci(num):

    if num == 0:      # Base condition

        return 0

    elif num == 1:    # Base condition

        return 1

    else:

        return (fibonacci(num - 1) + fibonacci(num - 2) )

f1 = fibonacci(6)
print(f1)






# Task 1(c)
# Print array elements



def array_elemnents(arr):

    if len(arr) > 1:      # Base condition

        data = arr[0]

    else:
        return arr[0]     # Base condition


    arr = arr[1:]
    return str(data) + ", " + str(array_elemnents(arr))


lst = [1, 2, 3, 4, 5]
print(array_elemnents(lst))






# Task 1(d)
# Exponent


def powerN(base , power):

    if power == 0:       # Base condition
        return 1

    x = powerN(base, power // 2)

    if power % 2 == 0:
        return x * x

    else:
        return x * x * base


p1 = powerN(3 , 2)
print(p1)








### Solution of Task 02 ###


# Task 2(a)
# Decimal to Binary converter


def DecimalToBinary(n):

    if n > 1:      # Base condition
        DecimalToBinary(n // 2)

    print(n % 2, end = '')

DecimalToBinary(10)






# Task 2(b,c)
# Addition and Reverse, both are in one class


class Node:
    def __init__(self, element, n = None):
        self.element = element
        self.next = n

class MyList:

    # Singly linked list constructor
    def __init__(self, array):

        self.head = None

        # ==== Linked list ====
        tail = None
        for data in array:
            itr = Node(data, None)

            if self.head is None:
                self.head = itr
                tail = itr

            else:
                tail.next = itr
                tail = itr

        # ======================

    # linked list print
    def showList(self):
        if self.head == None:
            print('Empty List')

        else:
            value = self.head

            while value != None:
                print(value.element, end = ' ')
                value = value.next





# Task 2(b)
# Addition

def add(var):

    Sum = 0

    if var == None:      # Base condition
        return 0

    Sum = Sum + var.element + add(var.next)
    return Sum





# Task 2(c)
# Reverse

def reverse(vari):

    if vari == None:      # Base condition
        return

    reverse(vari.next)    # reverse print
    print(vari.element, end = ' ')




# tester class of Task 2(b,c)
a = [1, 2, 3, 4 ]

arr = MyList(a)
print("Original Linked List")
arr.showList()
print()

print('\nSum of the linked list')
print(add(arr.head))
print()

print("Reversed List")
reverse(arr.head)
print()






### Solution of Task 03 ###



# Task 03
# Hoc Builder


def hocBuilder(height):

    if height == 0:    # Zero level bulding need no cards
        return 0

    if height == 1:    # Top level of bulding needs 8 cards
        return 8

    if height > 1:
        return 5 + hocBuilder(height - 1)

print(hocBuilder(3))






### Solution of Task 04 ###



# Task 4(a)
# Left side pattern


def pattern(row, column = 1):

    if row == 0:         # Base condition
        return           # no line left


    printline(column)
    print()

    pattern(row - 1, column + 1)

def printline(column):

    if column == 0:     # Base condition
        return

    printline(column - 1)
    print(column, end = ' ')

pattern(int(input('Input : ')))




# Task 4(b)
# Right side pattern


def pattern(row, column = 1):

    if row == 0:         # Base condition
        return           # no line left

    # print doubled space at the beginning
    print(' ' * (row - 1) * 2, end = '')


    printline(column)
    print()

    pattern(row - 1, column + 1)

def printline(column):

    if column == 0:     # Base condition
        return

    printline(column - 1)
    print(column, end = ' ')

pattern(int(input('Input : ')))






### Solution of Task 05 ###




# Task 5
# Investment

class FinalQ:
    def print(self,array,idx):

        if(idx < len(array)):
            profit = self.calcProfit(array[idx])

            # TO DO ============
            profit = "{:.1f}".format(profit)
            print(str(idx + 1) , ".  Investment: ", array[idx] , ";  Profit: ", profit)

            self.print(array, idx + 1)

        return

    def calcProfit(self, investment):

        # TO DO ============

        if investment <= 25000:
            return 0




        if investment <= 100000:

            # RecursionError: maximum recursion depth exceeded in comparison
            # Each 25000 USD gives a profit of 1125 dollar. Therefore, it's adding 1125 each time (upto 50000)
            # 25000 + 25000 = 50000 ; (25000*4.5)/100 = 1125
            # Error handled


            if investment >= 50000:
                return 1125 + self.calcProfit(investment - 25000)

            # The rest of dollers give a profit of 4.5% for each 100 USD (upto 25000)
            return 4.5 + self.calcProfit(investment - 100)




        # RecursionError: maximum recursion depth exceeded in comparison
        # Each 100000 USD gives a profit of 8000 dollar. Therefore, it's adding 8000 each time (upto 200000)
        # 100000 + 100000 = 200000 ; (100000*8)/100 = 8000
        # Error handled

        if investment >= 200000:
                return 8000 + self.calcProfit(investment - 100000)



        # The rest of dollers give a profit of 8% for each 100 USD (upto 100000)
        return 8 + self.calcProfit(investment - 100)



#Tester
array = [25000,100000,250000,350000]
f = FinalQ()
f.print(array, 0)
```
