### 1. Explique o conceito de uma Fila de Prioridade como um Tipo de Dados Abstrato e dê um exemplo prático de situação onde o uso desse TDA é vantajoso.

```python
'''
Fila, um TAD...

- Ordenado,
- do tipo FIFO (First In, First Out (Fila e FILO são bem mnemônicos)),

tem 3 operações principais e seu best case é casos que representem ordem de chegada
  (onde o primeiro a chegar vai receber o push e qndo começar X processo, comece o pop pelo 1st.)
'''

# Implementação

class Fila:
    def __init__(self): # construc
        self.items = []

    def esta_vazia(self): # para primeiro estado
        return self.items == []

    def enfileirar(self, item): # push, por
        self.items.append(item)

    def desenfileirar(self):
        if not self.esta_vazia(): # Pop
            return self.items.pop(0)
        else:
            return None
```

```python
# Recursivo - Função que pode chamar a si própria (acumula chamadas)

def recurs(entrada, passo=None): #none para a primeira vez

    if passo is None:
        passo = 0

    n = int(entrada)
    passo += 1  # Adiciona passo
    print(passo)

    if n == 1: # Recursividade propriamente dita
        return passo
    elif n % 2 == 0:
        return recurs(n // 2, passo)
    else:
        return recurs(3 * n + 1, passo)
```

### 3. Apresente e implemente (a linguagem fica a sua escolha) um problema clássico que é comumente resolvido com abordagem recursiva, como o algoritmo de Torres de Hanoi, e explique a estratégia por trás da solução recursiva.

```python
# função das torres, que conta os movimentos

def torre(disco):
    if disco == 1:
        return 1
    elif disco == 2:
        return 3
    else:
        return 1 + (2 * torre(disco - 1))
```

### 5. Implemente uma função em uma linguagem de sua escolha para inserir um nó em uma lista encadeada circular de pessoas (sugira os atributos), levando em consideração a ordenação por idade. O novo nó deve ser inserido na posição correta da lista para manter a ordem crescente de idade.

```python

# o formulário, a informação do node
class Pessoa:
    def __init__(self, nomeInput, idadeInput):
        self.nome = nomeInput
        self.idade = idadeInput
        self.proximo = None

# a classe pa criar a lista
class ListaCircular:
    def __init__(self):
        self.head = None

    def inserir_ordenado(self, nome, idade):
        novoNode = Pessoa(nome, idade)

        if self.head is None:  # Lista vazia
            self.head = novoNode
            self.head.proximo = self.head
        elif idade < self.head.idade:  # Inserir antes da cabeça
            novoNode.proximo = self.head
            atual = self.head
            while atual.proximo != self.head:
                atual = atual.proximo
            atual.proximo = novoNode
            self.head = novoNode
        else:  # Inserir em outra posição
            anterior = None
            atual = self.head
            while atual.proximo != self.head and atual.idade < idade:
                anterior = atual
                atual = atual.proximo

            # faz a troca
            novoNode.proximo = atual
            anterior.proximo = novoNode
```
