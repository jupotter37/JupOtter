# Recursion by Geeky Shows

A Function calling itself again and again to compute a value is referred to recursive Function or Recursion.

```python
def myfun():
    print("Aadil")
    
myfun()
```

The below written code is an example of Recursion and if we execute this code it will run upto 1000 times, because a recursion function can call itself only for 1000 times. 

### Global Keyword in a Recursion

```python
i = 0
def myfun():
    global i
    i = 1
    print("Aadil")
    myfun()

myfun() 
```

After Execution of this code (1000 times), this code will throw an error:-<br>
RecursionError: maximum recursion depth exceeded while calling a Python object

```python
i = 0
def myfun():
    global i
    i = i + 1
    print("My Function:", i) 
    myfun()

myfun() 
```

```python
# Agar hum recursion mai global keyword nhi use karenge toh function value ko increase kar hi karega. 

i = 0
def myfun():
    i = 0
    i = i + 1
    print("My Function:", i) 
    myfun()

myfun() 
```

# How to Check the limit of Recursion

```python
import sys
sys.getrecursionlimit() 
```

# How to Set Recursion Limit

```python
sys.setrecursionlimit(4000)  
print(sys.getrecursionlimit()) 
```
