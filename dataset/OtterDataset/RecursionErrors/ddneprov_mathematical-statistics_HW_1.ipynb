```python
def F(x):
    if x < 0:
        return 0
    else:
        return 1 - math.frexp(math.sqrt(x))[0]
```

```python
def iF(x):
    if x >= 1 or x <= 0:
        return 0
    else:
        return math.log2(1 - x) / -3
```

```python
def myF(count):
    np.random.seed() # инициализация 
    X = np.random.uniform(0,1,count) # случайные числа ( от, до , количество) распределение
    y = np.zeros([X.shape[0]])
    for i in range(X.shape[0]):
        y[i] = iF(X[i])
    return y
```

```python
def Check(count):
    y = myF(count)
    ym = y.mean()
    ys = y.std()**2 * y.shape[0] / (y.shape[0]  - 1)
    t = sc.stats.t.ppf(0.975, count - 1)
    lhv = ym - t * (math.sqrt(ys)/math.sqrt(count))
    rhv = ym + t * (math.sqrt(ys)/math.sqrt(count))
    E = 0.4809
    return E > lhv and E < rhv
```

```python
countTrueList = []
for i in range(20):
    countTrue = 0
    for j in range(1000):
        if Check(10 + i):
            countTrue += 1
    countTrueList.append(countTrue / 1000)
plt.plot([10 + i for i in range(20)], countTrueList)
```
