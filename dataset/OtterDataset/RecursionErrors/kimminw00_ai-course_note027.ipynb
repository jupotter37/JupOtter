# 다중상속

```python
class A:
    def __init__(self):
        print('A')

        
class B(A):
    def __init__(self):
        print('B')
        A.__init__(self)

        
class C(A):
    def __init__(self):
        print('C')
        A.__init__(self)
        
        
class D(B, C):
    def __init__(self):
        print('D')
        D.__init__(self)
```

## 상속의 경우 연산 순서를 바꿀 경우, 연산 순서가 바뀌면서 결과가 달라질 수 있으므로 조심해야 한다.

```python
class A:
    def __init__(self):
        print('A')

        
class B(A):
    def __init__(self):
        A.__init__(self) # A대신 super로 바꿀 수 있다.
        print('B')

        
class C(A):
    def __init__(self):
        A.__init__(self)
        print('C')

        
class D(B, C):
    def __init__(self):
        B.__init__(self)  # function 방식으로 호출 : 클래스명이 먼저 앞에 온다. & 호출할 때 self값을 넘겨줘야 한다.
        C.__init__(self)
        print('D')
        # B의 생성자를 다시 호출하므로 재사용성이 높다.
        # method 방식 : 인스턴스 뒤에 점을 찍어서 호출 & 호출할 때 self를 생략해도 된다.
```

```python
d = D()

# A의 생성자가 2번 호출되면서 Attribute가 2번 생성될 위험이 있음..!
# 한번씩만 호출하는 것이 목표
```

```python
vars(D)
# class의 경우 mapping proxy를 반환한다.
# method : self 를 생략해도 된다.
```

# [super](https://docs.python.org/3/library/functions.html?highlight=super#super)

# [guide to using super()](https://rhettinger.wordpress.com/2011/05/26/super-considered-super/)

> In a class hierarchy with single inheritance, super can be used to refer to parent classes 
>
> without naming them explicitly, thus making the code more maintainable.

> The second use case is to support cooperative multiple inheritance in a dynamic execution environment.


```python
class A:
    def __init__(self):
        # super().__init__() 이거 호출해야하나??????
        # super(A, self).__int__() 이렇게 써도 똑같다. 자명해서 생략
        print('A')

        
class B(A):
    def __init__(self):
        super().__init__() # super는 method 방식이다.
        print('B')

        
class C(A):
    def __init__(self):
        super().__init__()
        print('C')

        
class D(B, C):
    def __init__(self):
        super().__init__()
        print('D')        
```

Super는 method방식이다. (method는 첫번쨰 인자를 안쓴다.)
 vs
클래스명을 명시해서 __init__을 비교하는 경우

```python
d = D()
# 스택에 넣어서 꺼내면서 순서를 결정한다.
# 스택 앞 : [D, B, C, A] : stack 에 이렇게 들어감 : A -> C -> B -> D 순서로 호출함 (stack역순)

# 출력이 아까 예제와 다르다.
# super는 부모대신 부르지만, super 쓸때와 아닐 때 행동방식이 다르다. 
```

```python
D.__mro__
# method resolution order : 다중상속 실행 순서 알려줌 
# 스택 앞 : [D, B, C, A] : stack 에 이렇게 들어감
# 가득 차면 하나씩 꺼내면서 함수를 호출한다.
```
