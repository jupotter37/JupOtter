```python
tenure_churn_no = df1[df1.Churn=='No'].tenure
tenure_churn_yes = df1[df1.Churn=='Yes'].tenure

plt.xlabel("Tenure")
plt.ylabel("No. of Customers")
plt.title("Customer Churn Visualization (Tenure)")

plt.hist([tenure_churn_yes, tenure_churn_no], color=['green', 'red'], label=['Churn = Yes', 'Churn = No'])
plt.legend()
```

With this Visualization there is trend that the customers are leaving who have the shorter tenure.
The user with tenure 0 to 10 months have a high probablity to leave the telecome service. There is high probability of retaining customer when they have a huge tenure.

```python
mc_churn_no = df1[df1.Churn=='No'].MonthlyCharges
mc_churn_yes = df1[df1.Churn=='Yes'].MonthlyCharges

plt.xlabel("Month Charges")
plt.ylabel("No. of Customers")
plt.title("Customer Churn Visualization (Monthly Charges)")

plt.hist([mc_churn_yes, mc_churn_no], color=['green', 'red'], label=['Churn = Yes', 'Churn = No'])
plt.legend()
```

Here the monthly charges are negatively affecting the user in the range of from 60 to 100 alot.

Similarly, there can be an infernce that the 'No internet service' and 'No internet service' can be set to 'No' to noramailize the data.

```python
changes_cols = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
               'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
for col in changes_cols:
    df1.loc[:, col].replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
for col in df1:
    print(f'{col} {df1[col].unique()}')
```

All the 'bool' values are now converted to int64 0's and 1's

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService', 'Contract', 'PaymentMethod'])
df2.columns 
```

```python
df2.sample(4)
```

```python
df2.dtypes
```

```python
df2.replace({True : 1, False: 0}, inplace=True)
```

```python
df2.sample(3)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['MonthlyCharges', 'TotalCharges', 'tenure']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col} {df2[col].unique()}')
```

With this the dataframe is now clean and has all the values between 0 and 1. Homogenized will help to train the ML on this dataset

Now, Let's do a train and split of the dataframe

```python
x = df2.drop('Churn', axis='columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2, random_state=5)
```
