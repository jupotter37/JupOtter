```python
#recusion 

def add(a,b):
    # if a < 0 throw an exception
    
    if a < 0 : raise ValueError("a must be natural number") 
    if a == 0: return b
    else: return add(a-1, b+1)
    
```

```python
add(5,4)
```

```python
add(5.4,2)
```

```python
# add value of range 
#make recursive function
def addRange(numlist,start,end):
    if(start > end) :return 0
    else: return add(numlist[start],addRange(numlist,start+1,end))
```

```python
addRange(list,0,len(list)-1)
```

```python
# FIb and recursion 
def fib(n):
    if n == 0: return 0
    elif n == 1: return 1
    else: return fib(n-1) + fib(n-2)
```

```python
fib(5)
```

```python
fib(66)
```
