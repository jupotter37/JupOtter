```python
#         ?
def binary_search(lst,value,first,last):
    
    if first > last:
        return False
    else :
        mid = (first + last) // 2
        if lst[mid] == value:
            if mid > 0 and lst[mid-1] == value :
                return binary_search(lst,value,mid-1,mid)
            return mid
        elif lst[mid] < value:
            return binary_search(lst,value,mid+1,last)
        else :
            return binary_search(lst,value,first,mid-1)
        
    
    
```
