# Aggregation (Has- A relation)

- Aggregation is a concept in OOP that helps establish a relationship between two classes.
- Describes a "has-a" relationship between objects, means: one class owns the other class
- Where one class contains or is composed of one or more objects of another class.  
- Eg:
  - customer and address class
  - The customer has an address 
  - means the customer class owns the address class
- This relationship is less strict than composition, meaning Independent Lifetimes, The contained objects (parts) can exist independently of the container object (whole).

```python
# Eg:
class Customer:
    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address= address # address is a complex entity(home, city, pincode, state)
    
    def print_address(self):
        print(self.address.get_city(), self.address.pin, self.address.state)

    def edit_profile(self, new_name, new_city, new_pin, new_state):
        self.name= new_name
        self.address.edit(new_city, new_pin, new_state)
        
class Address:
    def __init__(self, city, pin, state):
        self.__city = city # we make the city a private variable
        self.pin = pin
        self.state = state
        
    def get_city(self):# getter method
        return self.__city

    def edit(self, new_city, new_pin, new_state):
        self.__city = new_city 
        self.pin = new_pin
        self.state = new_state
         
        
add1= Address("Gurgaon", 3045, "Haryana")       
cust = Customer("george", "male", add1)#  objects of another class.
cust.print_address()

cust.edit_profile("Mohit",' Mumbai', 3244, "Maharashtra") 

cust.print_address()
```

# Inheritance
- It allows a class to inherit properties and methods from another class.
-  This means that a new class (called the child or derived class) can use the attributes and behaviors of an existing class (called the parent or base class)
- Promoting code reuse and reducing redundancy

### Key Points of Inheritance:
- Parent Class: Also known as the base or superclass, it is the class whose properties and methods are inherited by another class.

- Child Class: Also known as the derived or subclass, it is the class that inherits from the parent class.

- Reusability: Inheritance allows for the reuse of code, making the development process more efficient.

- Extensibility: Child classes can add new attributes and methods or override existing ones from the parent class.

### Method overriding
- Method overriding happens when a child class (subclass) has a method with the same name as a method in its parent class (superclass).
-  The child class's method will replace or "override" the parent's method when you call it on an instance of the child class.

## Super Keyword
- Super keyword in Python is a powerful tool used to call methods from a parent class inside a child class. This is especially useful for extending or modifying the behavior of inherited methods.
- super cannot access variable
- super cannot access outside the class

```python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price= price
        self.brand = brand
        self.camera= camera
        
    def buy(self):
        print("Buying a phone")
        
class SmartPhone(Phone):
    
    def buy(self):
        print("Buying a smartphone")
        # syntax to call parent buy method
        super().buy()
        
s=SmartPhone(20000, "Apple", 13)
s.buy()
```

```python
# super=> constructor
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price= price
        self.brand = brand
        self.camera= camera
        
    def buy(self):
        print("Buying a phone")
        

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        print("Inside SmartPhone constructor")
        super().__init__(price, brand, camera) # self is not return
        # super().brand # super cannot access the variable
        self.os = os
        self.ram = ram
        print("Inside SmartPhone constructor")

        
s=SmartPhone(20000, "Apple", 13, "Android",2)
print(s.os)
# print(s.brand) # cannot access the variable
# s.super().buy() it is used inside the class 
s.buy()

```

## Type of Inheritance

- Multilevel Inheritance

```python
#Single level
class Parent:
    def __init__(self):
        self.num=100

class Child(Parent):

    def __init__(self):
        super().__init__()
        self.var=200
        
    def show(self):
        print(self.num)
        print(self.var)

son=Child()
son.show()
```

```python
# Multilevel inheritance
class Product:
    def review(self):
        print("Product customer review")
class Phone(Product):
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying a phone")
class SmartPhone(Phone):
    pass

s= SmartPhone(20000, "Apple", 12)
s.buy()
s.review()
    
```

```python
# HierarchicL Inheritance
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand= brand
        self.camera= camera
    def buy(self):
        print("Buying a phone")
class SmartPhone(Phone):
    pass
class FeaturePhone(Phone):
    pass

SmartPhone(1000,"Apple","13px").buy()
FeaturePhone(10,"Lava","1px").buy()
```

```python
# Multiple Inheritane
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class Product:
    def review(self):
        print ("Customer review")

class SmartPhone(Phone, Product):
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
s.review()
```

```python
# the diamond problem
# https://stackoverflow.com/questions/56361048/what-is-the-diamond-problem-in-python-and-why-its-not-appear-in-python2
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class Product: 
    def buy(self):
        print ("Product buy method")

# Method resolution order
class SmartPhone(Phone, Product):# the class which first gets priority Python resolve method resolution order (MRO).
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
# Both parent classes contain the buy method and which parent method is executed
# is decided by how the name is returned in the first
# class SmartPhone(Phone,Product): here phone class is Return first 
```

Multiple Inheritance: Can lead to complexity, especially if the parent classes have methods with the same name, as it needs to resolve which method to use (this is known as the Diamond Problem in some languages).

```python
# Find the solution 
class A:

    def m1(self):
        return 20

class B(A):

    def m1(self):
        return 30

    def m2(self):
        return 40

class C(B):
  
    def m2(self):
        return 20
obj1=A()
obj2=B()
obj3=C()
print(obj1.m1() + obj3.m1()+ obj3.m2())
```

```python
class A:

    def m1(self):
        return 20

class B(A):

    def m1(self):
        val=super().m1()+30
        return val

class C(B):
  
    def m1(self):# recursion occur
        val=self.m1()+20
        return val
obj=C()
print(obj.m1())
```

# Polymorphism 
- having multiple faces (the same thing acts differently for different situations)
- Polymorphism is a fundamental concept in Object-Oriented Programming (OOP) that allows objects to be treated as instances of their parent class rather than their actual class.
- The term "polymorphism" means "many forms" and refers to the ability of different classes to be treated as instances of the same class through inheritance.

### Concept in Polymorphism
- Method Overriding
- Method Overloading
- Operator Overloading

1. <b>  Method Overriding

2. <b> Method overloading
   
     - Method overloading is a concept where multiple methods in the same class have the same name
     - But different parameters (different type, number, or both)OR different behavior.
     - when we give 1 input it will execute the method that needs one parameter and if 2  inputs it will execute the method have two parameter.
     - It advantage is the code look more clean
     - It allows the same method to perform different tasks based on the arguments passed to it.
     -   Python does not support method overloading in the same way as some other programming languages like Java or C++.

```python
class shape:
    def area(self, radius): # area of circle
        return 3.14*radius*radius
        
    def area(self, l, b): # area of rectangle
        return l*b
s= shape()
s.area(2)
s.area(3, 4) # in python it not work 
```

in python method overloading not work directly

```python
# There are many other ways but 
# One of the ways is default parameter argument
class shape:
    def area(self, a, b=0):
        if b==0:
            return 3.14*a*a
        else:
            return a*b
s= shape()
print(s.area(2))
print(s.area(3, 4))
```

3. <b> Operator Overloading
- A feature in Object-Oriented Programming that allows you to define or change the behavior of operators (like +, -, *, etc.) for user-defined classes.
- This means you can use operators with objects of your classes in a way that's intuitive and natural.
- Key Points of Operator Overloading
    - Special Methods: To overload an operator, you define special methods in your class. These methods are also known as magic methods or dunder methods (because they start and end with double underscores).

    - Intuitive Syntax: Allows objects to be manipulated using standard operators, making the code more readable and expressive.
- Common Operator Overloading Methods in Python
    - __add__ for +

    - __sub__ for -

    - __mul__ for *

    - __truediv__ for /

    - __eq__ for ==

    - __lt__ for <

    - __gt__ for > 

- same operator behaviour is different for different input
