# Nested Function
A nested function is declared inside another function and can access variables from the enclosing (outer) function. This concept is useful for creating helper functions that are only relevant within the context of the outer function.

```python
def outer():
    def inner():
        print("This is inner function")
    inner()
    print("This is outer function")
```

```python
outer()
```

```python
inner()
```

```python
def outer():
    def inner():
        print("This is inner function")
        outer()
        
    inner()
    print("This is outer function")
```

```python
outer()
```
