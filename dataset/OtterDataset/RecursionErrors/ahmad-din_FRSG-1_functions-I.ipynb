```python
def gcd(m,n):
    fm = []
    for i in range(1,m+1):
        if(m%i)==0:
            fm.append(i)
    fn = []
    for j in range(1,n+1):
        if(n%j)==0:
            fn.append(j)
    cf = []
    for f in fm:
        if f in fn:
            cf.append(f)
    return (cf[-1])
```

```python
gcd(14,63)
```

```python
def findpos(l,v):
    for i in range(len(l)):
        if l[i]==v:
            pos = i
            break
        else:
            pos = -1
        return(pos)
l = [1,2,3,4,5,6]
findpos(l,3)
```

```python
def search(seq, v):
    for x in seq:
        if x==v:
            return True
    return False
seq = [1,2,3,4,5,6]
search(seq, 7)
```

```python
def bsearch(seq,v,l, r):
    if(r-l==1):
        return(False)
    mid = l+r//2 
    if(v==seq[mid]):
        return(True)
    elif (v<seq[mid]):
        return(bsearch(seq,v,l,mid))
    else:
        return(bsearch(seq,v,l,mid+1,r))
seq=[1,2,3,4,5,6,7]
l,r=0, len(seq)
bsearch(seq,4,l,r)
```

```python
def selectionsort(l):
    for start in range(len(l)):
        minpos = start
        for i in range(start, len(l)):
            if l[i]< l[minpos]:
                minpos = i
        (l[start],l[minpos])=(l[minpos], l[start])
l = [3,2,4,5,6,1]
print(l)
# for l in range(5,0,-1):
#     print(l)
selectionsort(l)
print(l)
```
