```python
# Selection sort [O(n^2)]
A = [4, 2, 9, 1, 0, -3, 18]
def selectionSort(arr):

    for i in range(len(arr)):
        min_i = i
        for j in range(i+1,len(arr)):
            if arr[j] < arr[min_i]:
                min_i = j
        arr[i], arr[min_i] = arr[min_i], arr[i]

selectionSort(A)
print(A)
```

```python
# Stable selection sort [O(n^2)]
A = [4, 2, 9, 1, 0, -3, 18]
def stableSelectionSort(arr):
    
    for i in range(len(arr)):
        min_i = i
        for j in reversed(range(i+1,len(arr))):
            if arr[j] < arr[min_i]:
                min_i = j
        arr[min_i], arr[i] = arr[i], arr[min_i]

stableSelectionSort(A)
print(A)
```

```python
# Merge sort [O(nlog(n)]
A = [4, 2, 9, 1, 0, -3, 18]
def merge(arr, lb, mid, ub):
    
    n1 = mid - lb + 1
    n2 = ub - mid

    # create temp arrays
    L = [0] * n1
    R = [0] * n2

	# Copy data to temp arrays L[] and R[]
    for i in range(n1):
        L[i] = arr[lb + i]

    for j in range(n2):
        R[j] = arr[mid + 1 + j]

    # Merge the temp arrays back into arr[l..r]
    i = 0   # Initial index of first subarray
    j = 0   # Initial index of second subarray
    k = lb  # Initial index of merged subarray

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    # Copy the remaining elements of L[], if there are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    # Copy the remaining elements of R[], if there are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1


def mergeSort(arr, lb, ub):
    if lb < ub:
        mid = (lb+ub)//2
        mergeSort(arr, lb, mid)
        mergeSort(arr, mid+1, ub)
        merge(arr, lb, mid, ub)

mergeSort(A, 0, len(A)-1)
print(A)
```

```python
# Counting sort [O(n)] - only for positive integers
A = [4, 2, 9, 1, 0, 3, 18]
def countingSort(arr):
    count_arr = [0] * (max(arr) + 1)
    for i in range(len(arr)):
        count_arr[arr[i]] += 1
    for i in range(1, len(count_arr)):
        count_arr[i] += count_arr[i-1]
    output = [0] * len(arr)
    for i in range(len(arr)):
        output[count_arr[arr[i]]-1] = arr[i]
        count_arr[arr[i]] -= 1
    return output

print(countingSort(A))
```

```python
# Quick sort average case: [O(nlog(n))], worst case: [O(n^2)]
A = [4, 2, 9, 1, 0, -3, 18]
def quickSort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr.pop()
    left, right = [], []
    for item in arr:
        if item < pivot:
            left.append(item)
        else:
            right.append(item)
    return quickSort(left) + [pivot] + quickSort(right)

print(quickSort(A))
```

```python
# Randomized quick sort average case: [O(nlog(n))], worst case: [O(n^2)]
import random
A = [4, 2, 9, 1, 0, -3, 18]

def quickSort(arr):
    if len(arr) <= 1:
        return arr
    random_int = random.randint(0, len(arr) - 1)
    pivot = arr.pop(random_int)
    left, right = [], []
    for item in arr:
        if item < pivot:
            left.append(item)
        else:
            right.append(item)
    return quickSort(left) + [pivot] + quickSort(right)

print(quickSort(A))
```

```python
def longest_common_substring(s1, s2):
   m = [[0] * (1 + len(s2)) for _ in range (1 + len(s1))]
   longest, x_longest = 0, 0
   for x in range(1, 1 + len(s1)):
       for y in range(1, 1 + len(s2)):
           if s1[x - 1] == s2[y - 1]:
               m[x][y] = m[x - 1][y - 1] + 1
               if m[x][y] > longest:
                   longest = m[x][y]
                   x_longest = x
           else:
               m[x][y] = 0
   return s1[x_longest - longest: x_longest]
```

```python
fib_cache = {0 : 0, 1 : 1}

def fib(n):
    if n in fib_cache:
        return fib_cache[n]
    for i in range(100, n, 100):
        fib(i)
    fib_cache[n] = result = fib(n-1) + fib(n-2)
    return result
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci(n):
    if n in {0, 1}:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
class Node:
    def __init__(self, unit, multiplier):
        self.unit = unit
        self.multiplier = multiplier
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def add_node(self, unit, multiplier):
        new_node = Node(unit, multiplier)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
    
    def print_list(self):
        current_node = self.head
        while current_node is not None:
            print(current_node.unit, current_node.multiplier)
            current_node = current_node.next
```
