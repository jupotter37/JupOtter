# Binary Search (p. 9)

```python
len(my_list) # length is 5, but index position is length - 1 bc python counts from zero
```

```python
def binary_search(list, item):
    low = 0 #low and high keeps track of which part of the list you're searching in
    high = len(list)-1 
    
    while low <= high: #while you haven't narrowed it down to one element ...
        mid = (low + high) // 2 #check the middle element (note: need to use floor division for python 3 (difference from book))
        guess = list[int(mid)]
        if guess == item: #found the item
            return mid
        if guess > item: #if the guess was too high
            high = mid - 1 #lower the upper limit to the mid - 1
        else: #else the guess was too low
            low = mid + 1 #raise the lower limit to the mid + 1
    return None
```

```python
my_list = [1, 3, 5, 7, 9]
```

```python
print(binary_search(my_list, 3)) #second slot has index 1, so it returns 1
```

```python
print(binary_search(my_list, -1)) #returns 'None' because -1 is not in our list so it didn't find it.
```

# Selection Sort (p. 36)

```python
def findSmallest(arr):
    smallest = arr[0] #stores the smallest value
    smallest_index = 0 #stores the index of the smallest value
    for i in range(1, len(arr)):
        if arr[i] < smallest: #compares the index value to the current value of smallest
            smallest = arr[i] #updates smallest if condition is true
            smallest_index = i #saves the index position of the smallest number
    return smallest_index #returns the index position of the smallest value
```

```python
def selectionSort(arr): #sorts an array
    newArr = []
    for i in range(len(arr)):
        smallest = findSmallest(arr) #finds the smallest element in the array, and adds it to the new array
        newArr.append(arr.pop(smallest))
    return newArr
```

```python
print(selectionSort([5, 3, 6, 2, 10]))
```

# Recursion (p. 38)

```python
def countdown(i):
    print(i)
    countdown(i-1)
```

```python
countdown(1) #this will run forever because we have not established a base and recursive case, so quit the script
```

So...when writing a recursive function, you need to tell it where to stop:

* Recursive case: when the function calls itself
* Base case: when the function doesn't call itself again (where the recursion stops)

so now we can we write our function like this:

```python
def countdown(i):
    print(i)
    if i <= 1: #base case
        return #ends the function if i is less than or equal to 1
    else: #recursive case
        countdown(i-1)
```

```python
countdown(20)
```

# Quicksort (p. 65)

```python
def quicksort(array):
    if len(array) < 2:
        return array # base case: arrays with 0 or 1 element are already sorted
    else:
        pivot = array[0] #recursive case
        less = [i for i in array[1:] if i <= pivot] # sub-array of all the elements less than pivot
        
        greater = [i for i in array[1:] if i > pivot] #sub-array of all the elements greater than the pivot 
        
        return quicksort(less) + [pivot] + quicksort(greater)
```

```python
print (quicksort([10, 5, 2, 3]))
```

# Implementing the breadth-first search algorithm (p. 110)

```python
def person_is_seller(name):
    return name[-1] == 'b'
```

```python
def search(name):
    search_queue = deque()
    search_queue += graph[name]
    searched = []
    while search_queue:
        person = search_queue.popleft()
        if not person in searched:
            if person_is_seller(person):
                print (person + " is a mango seller!")
                return True
            else:
                search_queue += graph[person]
                searched.append(person)
    return False
```

```python
search("you")
```
