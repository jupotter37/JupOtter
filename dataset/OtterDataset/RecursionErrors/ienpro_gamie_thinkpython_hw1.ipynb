```python
spam = input("문자를 입력하시오")

def do_twice(f):
    f()
    f()
def print_spam():

    print(spam)
    print('do_twice(print_'+spam+') :')

    do_twice(print_spam)

def do_four(g):

    do_twice(g)

    do_twice(g)
print('do_four(print_'+spam+') :')

do_four(print_spam)

print()
```
