```python
# recursion

def func(count):
  if count > 0:
    print(count, '현재')
    func(count-1)
  print('결과',count)
func(10)
```

```python
# 중첩함수: 함수 안에 정의된 함수

def func1(a):
  def func2():
    nonlocal a
    a= a+1
    return a
  return func2()
func1(2)
```

lambda 함수란 이름이 없는 한 줄 짜리 함수. 아래와 같이 작성
 - lambda 인수 나열 : 반환할 내용. 

인수가 없으면 생략 가능.

기억 하시나요? 함수의 인수로 함수도 받을 수 있음!

```python
def run(func, x):
  print(func(x))
run(lambda x: x+1, 1)
```

```python
# 유용한 내장함수

# zip
list1 = [1,2,3,4]
list2 = [5,6,7,8]
list(zip(list1, list2))
```

```python
# map(함수, 컬렉션)

list(map(lambda x: x+1, [1,2,3,4]))

# 함수와 컬렉션(list, tuple 등)을 매개변수로 받아 컬렉션의 모든 데이터를 함수의 매개변수로 대입해서 결과를 리턴하는 함수.
```

```python
# filter(함수, 컬렉션)

list(filter(lambda x: x%2 == 0, [1,2,3,4]))

# 함수와 컬렉션(list, tuple 등)을 매개변수로 받아 컬렉션의 모든 데이터를 함수의 매개변수로 대입해서 결과가 참인 컬렉션의 멤버들을 반환하는 함수
```
