## **4 : Calcul de a<sup>n</sup>**

Nous voulons créer une fonction qui calcule a<sup>n</sup> sans utiliser la fonction puissance.

Cet exemple est basé sur le fait que : 

**a<sup>n</sup>=a si n=1**

**a<sup>n</sup>=a x a<sup>n-1</sup> sinon** 

Ci-dessous, la fonction puiss(a,n) qui renvoie la valeur de a<sup>n</sup>. On a donc la transcription immédiate en Python :




````python
def puiss(a : int, n : int) -> int: 
    '''
    Cette fonction renvoie a**n sans utiliser **
    '''
    if n == 1 :
        return a
    else : 
        return a*puiss(a,n - 1) # la fonction s'appelle elle-meme
print(puiss(3,4))
````

Regardons ce qu'il se passe si on appelle puiss(3,4)

De façon triviale puiss(3,1) renvoie 3.

puiss(3,4) = 3 x puiss(3,3) = 3 x (3 x puiss(3,2)) = 3 x (3 x (3 x puiss(3,1))) = 3 x (3 x (3 x 3)) = 81

# **Fonctions récursives**



## **1 : Qu'est ce qu'une fonction récursive ?**

La récursivité est une méthode de résolution de problèmes qui consiste à décomposer le problème en sous-problèmes identiques de plus en plus simples jusqu'à obtenir un problème suffisamment simple qui puisse être résolu de manière évidente (triviale).

On peut appeler ce problème très simple : cas de base, ou cas trivial.

Dans l'exemple précédant le cas de base est : puiss(a,1) qui renvoie a.

## **2 : Pourquoi utiliser une fonction récursive ?**

Pour le calcul de a<sup>n</sup> , nous aurions pu tout simplement écrire le code suivant, qui est un code dit "itératif", sans utiliser une fonction récursive :

````python
def puiss_iter(a : int, n : int) -> int: 
    '''
    Cette fonction renvoie a**n sans utiliser **
    '''
    res = 1
    for i in range(n):
        res = res*a
    return res
print(puiss_iter(3,4))
````

Le plus souvent nous utilisons une fonction récursive, lorsque cela est beaucoup plus simple à programmer qu'un programme itératif (décomposition en sous-problèmes identiques de plus en plus simples).

## **3 : Conditions d'arrêt ?**



**Attention !** Une fonction récursive étant une fonction qui s'appelle elle-même, on peut créer une fonction qui "tourne" indéfiniment. 

Testez le code suivant :

````python
def puiss_fausse(a : int, n : int) -> int: 
    '''
    Cette fonction renvoie a**n sans utiliser **
    '''
    return a*puiss_fausse(a,n - 1) # la fonction s'appelle elle-meme
puiss_fausse(3,4)
````

Python s'arrête au bout d'un certain nombre d'appels récursifs, et affiche le message d'erreur. Il manquait dans cette fonction le cas de base :



```
if n == 1 :
       return a
```



Python limite explicitement le nombre d’appels récursifs dans une fonction. Ainsi, après 1000 appels récursifs, l’interpréteur Python va lever l’exception RecursionError et afficher le message d’erreur suivant : RecursionError: maximum recursion depth exceeded.

Cette limite, fixée à 1000 appels récursifs, est une valeur par défaut qu’il est possible de modifier en utilisant la fonction setrecursionlimit disponible dans le module sys. Par exemple, pour passer cette limite à 2000 appels maximum, on exécutera le code Python suivant :

       import sys
       sys.setrecursionlimit(2000)  

Un tel changement reste cependant dérisoire lorsque l’on a une définition récursive qui, par nature, effectue un très grand nombre d’appels emboîtés.


Il peut y avoir un autre problème si les appels récursifs successifs éloignent du cas de base au lieu d'en rapprocher.

Reprenons la fonction puiss() écrite plus haut de façon récursive. 

````python
def puiss(a : int, n : int) -> int: 
    '''
    Cette fonction renvoie a**n sans utiliser **
    '''
    if n == 1 :
        return a
    else : 
        return a*puiss(a,n - 1) # la fonction s'appelle elle-même
print(puiss(4,-1))
````

Ici puiss(4,-1) appelle puiss(4,-2), qui appelle puis(4,-3) etc...

On n'atteindra jamais le cas de base qui aurait été puiss(4,1), et le processus est infini

## **4 : Comment concevoir un programme récursif ?**

    

1.   Trouver le cas général : trouver l'élément de récursivité qui permet de décomposer le problème en cas plus simple. 
2.    Trouver le cas de base : trouver la condition d'arrêt, et la solution du problème dans ce cas particulier. La condition d'arrêt sera bien atteinte après un nombre fini d'appels. Chaque appel récursif tend à se rapprocher du cas de base.
3.     Réunir ces deux cas dans le programme : Le programme ne contient aucune boucle (for ou while) dans la partie qui résout le problème. Le programme contient un général une structure if / else introduisant le cas de base


   
  

Alice a réalisé le code suivant pour savoir si un entier se trouve dans un tableau (typ list python). C'est une recherche par dichotomie utilisant une fonction récursive. 

````python
def recherche(lst, x):
    '''
    Cette fonction renvoie True si x est dans lst, et False sinon
    Précondition : x : int, lst : list et ... (à vous de compléter !)
    Postcondition : cette fonction renvoie du type booléen
    '''
    if len(lst) == 0:
        return False
    
    m = len(lst) // 2

    if lst[m] == x:
        return True
    elif lst[m] < x:
        return recherche(lst[m + 1:], x)
    else:
        return recherche(lst[:m], x)
print(recherche([1,2,3,4,5,6],1))
print(recherche([1,2,3,4,5,6],6))
print(recherche([1,2,3,4,5,6],3))
print(recherche([1,2,3,4,5,6],8))  
print(recherche([1,2,3,4,5,6],0))
````

Dans tous les algorithmes que vous avez rencontrés, certaines techniques reviennent souvent. Notamment l'utilisation de :

Par exemple, calculer la moyenne des valeurs paires d'une liste demande un compteur (pour compter les valeurs paires) et un accumulateur (pour additionner les valeurs paires), et on renvoie le rapport somme des valeurs / nombre de valeurs

### **1 : Compteur récursif**

Reprenons un exemple de Compteur en impératif : compter le nombre de valeurs paires d'une liste. 

````python
def nbrePairs(l) :
    nb=0
    for element in l:
      if element%2==0:
            nb+=1
    return nb
print(nbrePairs([1,2,3,4,6]))
````

**Comment faire la même chose en mode récursif**

Notre compteur nb dans la fonction nbrePairs(l) peut se concevoir de façon récursive comme :

**nb = (1 si l[0] est pair, 0 sinon) + nbrePairs du reste**

Si le 1er nombre est pair on écrit nb = 1 + nombre pair du reste

Si le 1er nombre est impair on écrit nb = 0 + nombre pair du reste


remarquons que (1-l[0]%2) vaut 1 si l[0] est pair et 0 si l[0] est impair, et reprenons :


**nb = (1-l[0]%2) + nbrePairs du reste**= etc... Par exemple : nb=nbPairs([1,2,3,4,6])

nb=0 + nbPairs([2,3,4,6])................................. # 0 car 1 est impair

nb=0 + (1+nbPairs(3,4,6]) )...............................# 1 car 2 est pair

nb=0 + (1+ ( 0 + nbPairs([4,6]) )........................# 0 car 3 est impair

nb=0 + (1+ ( 0 + 1 nbPairs([6]) ) ) ) ................. # 1 car 4 est pair

nb=0 + (1+ ( 0 + ( 1 + ( 1 + ( nbPairs([]) ) ) ) ).. # 1 car 6 est pair

Et quand il n'y a aucun élément (liste vide) : nbPairs([]) = 0



Exemple : Un exemple récursif d'utilisation d'un Compteur. nb est le compteur.

````python
def nbrePairs(l) :
    if len(l) == 0 :
        return 0
    else :
        return 1-l[0]%2 + nbrePairs(l[1:])

print(nbrePairs([1,2,3,4,6]))
````

Vous remarquerez la simplicité de la traduction codée en python de la définition récursive :
nb = (1-l[0]%2) + nbpair du reste s'écrit très simplement :

    return 1-l[0]%2 + nbrePairs(l[1:])

De même, la condition d'arrêt : si l est vide, nb=0, s'écrit :

    if len(l) == 0 :
      return 0

### **2 : Accumulateur récursif**

> Bloc en retrait



Reprenons un exemple d'accumulateur en impératif : calculer la somme des éléments d'une liste :

````python
def somme(l) :
    s=0
    for element in l:
        s+=element
    return s
print(somme([1,2,3,4,6]))
````

Notre fonction somme(l) ci-dessus se comprend de façon récursive comme :

S= l[0] + somme du reste = l[0] + ( l[1] + somme du reste ) = etc...

jusqu'à ce que le reste ne contienne aucun élément (liste vide) et la somme sera alors 0.

Exemple :

s=somme([2,4,1,5])

= 2 + somme([4,1,5])

= 2 + (4 + somme([1,5]) )

= 2 + (4 + (1 + somme([5]) ) )

= 2 + (4 + (1 + (5 + somme([]) ) ) )

et somme([]) vaut 0, donc s = 2 + (4 + (1 + (5 + 0) ) ) ) = 10

**Exercice de code : Un exemple d'utilisation d'un Accumulateur en récursion: s est l'accumulateur.**

````python
def somme(l) :
  '''
  renvoie la somme des éléments de la liste (supposés être des nombres)
  >>> somme('1','6',8')
  15
  '''

  if len(l) == 0:
    return 0
  else:
    return l[0] + somme(l[1:])
      
  # vous devez coder la fonction somme en mode récursif

print(somme([1,2,3,4,6]))
````

### **3 : Curseur récursif**

**Recherche du max : on déplace le curseur en notant la valeur**

Envisageons d'abord le cas plus simple de la recherche de la plus grande valeurs dans une liste de nombres : l=[1, ,4, 3, 7, 6] le max est 7.

Nous avons déjà travaillé cet algorithme : on initialise vmax=1, puis on compare vmax au suivant, si le suivant est plus grand on actualise vmax, puis on continue jusqu'au dernier élément.

Si on pense récursif ici cela donne :

l=[1, ,4, 3, 7, 6]

vmax = max(l)

= max(1, max [4,3,7,6])

= max(1 ,max (4, max([3,7,6] ) )

= max(1 ,max (4, max(3,max[7,6] ) ) )

= max(1 ,max (4, max(3,max(7,max[6]) ) ) )

En parvenant à la dernière ligne, on cherche le max[6], qui est trivialement égal à 6, on arrête donc les appels. (quand la liste n'a qu'un élément on renvoi cet élément)

````python
def getMax(l : list) -> float :
    ''''
    recherche récursive de la valeur max dans une liste de float
   '''
    if len(l) == 1:
        return l[0]
    else :
       vmaxReste= getMax(l[1:])
       vmax=l[0] if l[0] >vmaxReste else vmaxReste
       return vmax
print(getMax([1,2,5,2,1]))
````

**Cas de la recherche de l'indice du max : on déplace le curseur en notant l'indice**

Dans le cas ou l'on recherche l'indice du max de façon récursive, nous allons devoir bien comprendre un point plus délicat.

Prenons un exemple : l = [ 1, 2, 4, 3 ]

on va considérer que la position du max est :


0 ou (1+ position du max dans [2,4,3])


en effet, dans notre exemple, l'indice du max dans [2,4,3] est 1, mais cela correspond à l'indice 2 dans l.

Décomposons complètement un exemple :

l=[1,2,5,2] l'indice du max est soit 0 soit (1 + indice du max dans [2,5,2])
puis indice du max de [2,5,2] = 0 ou (1 + indice du max de [5,2])

Et indice du max de [5,2] = 0 ou (1+indice du max dans [2] ) mais cette dernière liste n'ayant qu'un seul élément, le max est forcément à l'indice 0.

Ce qu'il faut bien comprend c'est le 1 + indice du max du reste.
La fonction s'écrira donc :

````python
def idxMax(l) :
    if len(l)==1 :
        return 0
    else :
        idxMaxReste=1+idxMax(l[1:])
        if l[0]>l[idxMaxReste]:
          idx=0  
        else:
          idx=idxMaxReste
        return idx
print(idxMax([1,2,3,4,2]))
````

la ligne clé est :

        idxMaxReste=1+idxMax(l[1:])

L'utilisation du curseur est donc un peu plus compliquée à bien comprendre. 

En fait, idx est l'indice du max dans la liste, mais dans la fonction appelante, qui contient un élément de plus en tête, l'indice est idx+1. 

Une fois bien compris ceci, l'utilisation de curseur ne devrait plus vous poser de difficulté.

La condition d'arrêt si len(l)==1 : return 0 est intuitive
