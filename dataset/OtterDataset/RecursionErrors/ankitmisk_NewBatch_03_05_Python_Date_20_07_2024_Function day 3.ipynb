```python
def rec_factorial(n):
    if n == 1:
        return n
    else:
        return rec_factorial(n-1)*n
```

```python
rec_factorial(5)
```

```python
# 5*4*3*2*1
```

```python
def rec_sum(n):
    if n == 1:
        return n
    else:
        return rec_sum(n-1)+n
```

```python
# decorator:- decorate already existing func or modify
```

```python
def Beauty_parlour(koi_func):
    print('Now you are Decorated')
    koi_func()
    print('💥💥')
    
```

```python
def jatin():
    print('I am Jatin')
    print('Let\'s decorate me')
```

```python
jatin()
```

```python
Beauty_parlour(jatin)
```

```python
@Beauty_parlour
def mohan():
    print('I am Mohan')
    print('Let\'s decorate me')
    
@Beauty_parlour
def jatin():
    print('I am Jatin')
    print('Let\'s decorate me')
```

```python
def even(data):
    result = []
    for i in data:
        if i % 2 == 0:
            result.append(i)
    return result
```

```python
id(even)
```

```python
numbers = [23,4,23,45,54,1243,567,543,46]

```

```python
all_sales = [23,4,12,4356,5,4532,5467,6,43,54,67,4,6]


(lambda num: num if num % 3 == 0 else None)
```
