```python
def fact(n):
    if n <= 1:
        return 1
    return n*fact(n-1)        
```

```python
fact(10)
```

Quiz. 1부터 n까지의 합 구하기를 재귀호출로 만들어 보시오

```python
def plus(n):
    if n <= 1:
        return 1
    return n+plus(n-1)
```

```python
plus(100)
```

숫자 n개 중에서 최댓값 찾기를 재귀호출로 만들어 보시오

```python
def find_max(a,n):
    if n == 1:
        return a[0]
    max_n_1 = find_max(a,n-1)
    if max_n_1>a[n-1]:
        return max_n_1
    else:
        return a[n-1]
```

```python
l=[999,1,2,6,7,5,9999999,3,8,6,9,4,10,4,654,54,984,5,58,64,64545131]

find_max(l, len(l))
```
