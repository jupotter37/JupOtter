```python
#merge sort
def merge(low,mid,high):
    left=low
    right=mid+1
    res=[]
    while left<=mid and right<=high:
        if nums[left]<=nums[right]:
            res.append(nums[left])
            left+=1
        else:
            res.append(nums[right])
            right+=1
    while left<=mid:
        res.append(nums[left])
        left+=1
    while right<=high:
        res.append(nums[right])
        right+=1
    for i in range(low,high+1):
        nums[i]=res[i-low]

def mergeSort(low,high):
    if low>=high:
        return
    mid=(low+high)//2
    mergeSort(low,mid)
    mergeSort(mid+1,high)
    merge(low,mid,high)

nums=list(map(int,input("enter the input ").split()))
mergeSort(0,len(nums)-1)
print(nums)
```

```python
def permute(nums):
    print("nums",nums)
    if len(nums)==1:
        return [nums[:]]
    res=[]
    for i in range(len(nums)):
        first=nums.pop(0)
        perms=permute(nums)
        for p in perms:
            p.append(first)
        res.extend(perms)
        nums.append(first)
    print("res",res)
    return res
print(permute([1,2,3]))
```

```python
#quick sort
def partition(low,high):
    pivot=nums[low]
    i,j=low,high
    while i<j:
        while nums[i]<=pivot and i<high:
            i+=1
        while nums[j]>pivot and j>low:
            j-=1
        if i<j:
            nums[i],nums[j]=nums[j],nums[i]
    nums[low],nums[j]=nums[j],nums[low]
    return j

def quicksort(low,high):
    if low<high:
        mid=partition(low,high)
        quicksort(low,mid-1)
        quicksort(mid+1,high)

nums=list(map(int,input("enter the numbers").split()))
quicksort(0,len(nums)-1)
print(nums)
```
