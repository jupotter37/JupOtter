```python
def factorial(n):
    ans = 1
    for i in range(1,n+1):
        ans *= i
    return ans
```

```python
factorial(5)
```

```python
def rec_factorial(n):
    if n == 1:
        return n
    else:
        return rec_factorial(n-1)*n
```

```python
def even_odd(data):
    even = []
    odd = []
    for i in data:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)        
    print('Odd list',odd)
    print('even list',even)    
```

```python
even_odd([23,2345,54,31234,6,77,5354])
```

```python
mera_data = [23,2345,54,31234,6,77,5354]
```

```python
ek = lambda x: x if x % 2 == 0 else None
```

```python
print(ek(20))
```

```python
mera_data
```

```python
list(filter(ek,mera_data))
```

```python
list(filter(lambda x: x if x % 2 == 0 else None,mera_data))
```

```python
list(filter(lambda x: x if x % 2 != 0 else None,mera_data))
```

```python
print('Even list: ',list(filter(lambda x: x if x % 2 == 0 else None,mera_data)))
print('Odd list: ',list(filter(lambda x: x if x % 2 != 0 else None,mera_data)))
```
