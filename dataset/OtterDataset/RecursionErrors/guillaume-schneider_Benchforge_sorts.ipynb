# Analyse de la Complexité Théorique des Algorithmes de Tri

---
## Tri à Bulle (Bubble Sort)
Le tri à bulle compare deux éléments adjacents et les échange s'ils ne sont pas dans le bon ordre.

### Complexité Théorique

- **Pire cas** : $O(n^2)$
   - Si le tableau est trié dans l'ordre inverse, chaque élément doit être comparé et échangé à chaque itération.

- **Meilleur cas** : $O(n)$
   - Si le tableau est déjà trié, une optimisation (en arrêtant dès qu'il n'y a plus d'échanges) permet de réduire la complexité.

- **Cas moyen** : $O(n^2)$
   - En général, chaque élément est comparé avec plusieurs autres dans un tableau d'entrée non trié.

### Pseudocode

```plaintext
BubbleSort(array)
    n ← longueur de array
    pour i de 0 à n-1
        pour j de 0 à n-i-2
            si array[j] > array[j+1]
                échanger array[j] et array[j+1]

````python
result = benchmark.test(SortType.BUBBLE_SORT)
````

---
## Tri par Sélection (Selection Sort)
Le tri par sélection sélectionne à chaque itération le plus petit (ou le plus grand) élément et le place dans la position correcte.

### Complexité Théorique
- **Pire cas** : $O(n^2)$
   - Chaque élément est comparé à tous les autres pour trouver le minimum.

- **Meilleur cas** : $O(n^2)$
   - Même si le tableau est déjà trié, l'algorithme effectue toutes les comparaisons.

- **Cas moyen** : $O(n^2)$
   - Le nombre de comparaisons est toujours le même, quel que soit l'état initial.

### Pseudocode

```plaintext
SelectionSort(array)
    n ← longueur de array
    pour i de 0 à n-1
        minIndex ← i
        pour j de i+1 à n-1
            si array[j] < array[minIndex]
                minIndex ← j
        échanger array[i] et array[minIndex]

````python
benchmark.test(SortType.SELECTION_SORT)
````

---
## Tri par Insertion (Insertion Sort)
Le tri par insertion insère chaque élément dans la partie triée du tableau.

### Complexité Théorique
- **Pire cas** : $O(n^2)$
   - Si le tableau est trié dans l'ordre inverse, chaque élément doit être déplacé sur toutes les positions précédentes.

- **Meilleur cas** : $O(n)$
   - Si le tableau est déjà trié, l'algorithme effectue seulement \( n-1 \) comparaisons.

- **Cas moyen** : $O(n^2)$
   - Les déplacements et les comparaisons augmentent quadratiquement en fonction de la taille du tableau.


### Pseudocode

```plaintext
InsertionSort(array)
    n ← longueur de array
    pour i de 1 à n-1
        clé ← array[i]
        j ← i-1
        tant que j ≥ 0 et array[j] > clé
            array[j+1] ← array[j]
            j ← j-1
        array[j+1] ← clé

````python
benchmark.test(SortType.INSERTION_SORT)
````

## Tri par Tas (Heap Sort)
Le tri par tas utilise une structure de données (tas binaire) pour organiser les éléments avant de les trier.

### Complexité Théorique
- **Pire cas** : $O(n \log n)$
   - La construction et l'extraction du tas nécessitent $O(n \log n)$ opérations pour chaque élément.

- **Meilleur cas** : $O(n \log n)$
   - Même si le tableau est déjà trié, le processus d'organisation du tas reste identique.

- **Cas moyen** : $O(n \log n)$
   - Les opérations de tas sont constantes quel que soit l'état initial.

### Pseudocode

```plaintext
HeapSort(array)
    Construire un tas max à partir de array
    pour i de longueur de array - 1 à 0
        échanger array[0] et array[i]
        Réorganiser le tas (heapify) pour array[0] à array[i-1]

Heapify(array, n, i)
    largest ← i
    gauche ← 2*i + 1
    droite ← 2*i + 2
    si gauche < n et array[gauche] > array[largest]
        largest ← gauche
    si droite < n et array[droite] > array[largest]
        largest ← droite
    si largest ≠ i
        échanger array[i] et array[largest]
        Heapify(array, n, largest)

````python
benchmark.test(SortType.HEAP_SORT)
````

## Tri Rapide (Quick Sort)
Le tri rapide (Quick Sort) est un algorithme de tri efficace qui divise récursivement un tableau en partitionnant les éléments autour d'un pivot pour placer les plus petits à gauche et les plus grands à droite.

### Complexité Théorique
- **Pire cas** : $O(n^2)$
   - Se produit lorsque le pivot choisi divise mal le tableau, comme dans un tableau déjà trié.

- **Meilleur cas** : $O(n \log n)$
   - Chaque partition divise le tableau en deux moitiés égales.

- **Cas moyen** : $O(n \log n)$
   - En général, les partitions sont raisonnablement équilibrées.


### Pseudocode

```plaintext
QuickSort(array, low, high)
    si low < high
        pivotIndex ← Partition(array, low, high)
        QuickSort(array, low, pivotIndex-1)
        QuickSort(array, pivotIndex+1, high)

Partition(array, low, high)
    pivot ← array[high]
    i ← low - 1
    pour j de low à high-1
        si array[j] ≤ pivot
            i ← i+1
            échanger array[i] et array[j]
    échanger array[i+1] et array[high]
    retourner i+1

````python
benchmark.test(SortType.QUICK_SORT)
````

## Tri Fusion (Merge Sort)
Le tri fusion divise le tableau en sous-parties, les trie, puis les fusionne.

### Complexité Théorique
- **Pire cas** : $O(n \log n)$
   - Les comparaisons et les fusions sont nécessaires pour chaque division.

- **Meilleur cas** : $O(n \log n)$
   - Même pour un tableau déjà trié, les sous-tableaux sont fusionnés.

- **Cas moyen** : $O(n \log n)$
   - Les opérations restent constantes quel que soit l'état initial.

### Pseudocode

```plaintext
MergeSort(array, gauche, droite)
    si gauche < droite
        milieu ← (gauche + droite) // 2
        MergeSort(array, gauche, milieu)
        MergeSort(array, milieu+1, droite)
        Fusionner(array, gauche, milieu, droite)

Fusionner(array, gauche, milieu, droite)
    n1 ← milieu - gauche + 1
    n2 ← droite - milieu
    créer tableaux gaucheArray et droiteArray
    pour i de 0 à n1-1
        gaucheArray[i] ← array[gauche + i]
    pour j de 0 à n2-1
        droiteArray[j] ← array[milieu + 1 + j]
    i, j, k ← 0, 0, gauche
    tant que i < n1 et j < n2
        si gaucheArray[i] ≤ droiteArray[j]
            array[k] ← gaucheArray[i]
            i ← i + 1
        sinon
            array[k] ← droiteArray[j]
            j ← j + 1
        k ← k + 1
    copier les éléments restants de gaucheArray et droiteArray dans array

````python
benchmark.test(SortType.MERGE_SORT)
````

# Analyse des Algorithmes de Tri

---

## Comparaison des Algorithmes

| Algorithme       | Meilleur cas      | Cas moyen       | Pire cas        |
|------------------|-------------------|-----------------|-----------------|
| Bubble Sort      | $O(n)$            | $O(n²)$         | $O(n²)$         |
| Selection Sort   | $O(n²)$           | $O(n²)$         | $O(n²)$         |
| Insertion Sort   | $O(n)$            | $O(n²)$         | $O(n²)$         |
| Heap Sort        | $O(n \log n)$      | $O(n \log n)$    | $O(n \log n)$    |
| Quick Sort       | $O(n \log n)$      | $O(n \log n)$    | $O(n²)$         |
| Merge Sort       | $O(n \log n)$      | $O(n \log n)$    | $O(n \log n)$    |

---

### **Bubble Sort**
- **Performance :**
  - Très inefficace pour les grands tableaux en raison du nombre élevé de comparaisons et permutations.
  - Particulièrement lent sur des tableaux inversés.
- **Points forts :**
  - Simple à implémenter.
- **Limites :**
  - Inefficace sur les grands tableaux.

---

### **Selection Sort**
- **Performance :**
  - Nombre constant de permutations $n-1$, mais un grand nombre de comparaisons.
  - Peu impacté par l'ordre initial des données.
- **Points forts :**
  - Stable en termes de permutations.
- **Limites :**
  - Lent par rapport aux algorithmes modernes.

---

### **Insertion Sort**
- **Performance :**
  - Plus rapide que Bubble et Selection pour des tableaux de petite ou moyenne taille.
  - Optimal sur des données presque triées.
- **Points forts :**
  - Efficace pour de petits tableaux.
  - Pratique pour des flux de données en temps réel.
- **Limites :**
  - Inefficace sur des tableaux désordonnés de grande taille.

---

### **Heap Sort**
- **Performance :**
  - Stable dans toutes les situations grâce à sa structure arborescente.
  - Plus lent que Quick Sort en pratique, mais plus fiable.
- **Points forts :**
  - Stable et adapté aux gros volumes de données.
- **Limites :**
  - Complexe à implémenter comparé aux autres algorithmes.

---

### **Quick Sort**
- **Performance :**
  - Très rapide grâce à des constantes faibles.
  - Le choix du pivot est crucial.
- **Points forts :**
  - Adapté aux tableaux de grande taille.
- **Limites :**
  - Performances dégradées pour des tableaux triés ou presque triés.
  - Non stable par défaut.

---

### **Merge Sort**
- **Performance :**
  - Très régulier et efficace pour des grandes quantités de données.
  - Nécessite de la mémoire supplémentaire.
- **Points forts :**
  - Stable et adapté aux gros volumes de données.
- **Limites :**
  - Consomme $O(n)$ en mémoire additionnelle.

---

## Explication des Résultats

### Effet de la Structure Initiale

La disposition initiale des données joue un rôle majeur dans l’efficacité des algorithmes simples comme le **Bubble Sort**, le **Selection Sort** et l’**Insertion Sort**. Ces algorithmes sont très sensibles à l’ordre des données. Par exemple, le **Bubble Sort** peut détecter rapidement un tableau déjà trié, ce qui lui permet de réduire ses comparaisons et permutations, atteignant ainsi sa complexité minimale de $O(n)$. En revanche, sur un tableau inversé, il doit effectuer un nombre maximal d’opérations, atteignant $O(n^2)$.

L’**Insertion Sort** se distingue par ses performances sur des tableaux presque triés. Dans ce cas, il fonctionne de manière très efficace, car les déplacements nécessaires pour insérer un élément à sa position correcte sont minimaux. À l’inverse, le **Selection Sort** effectue toujours un nombre fixe de comparaisons, indépendamment de l’ordre initial, ce qui le rend moins flexible.

Les algorithmes avancés, comme le **Heap Sort**, le **Quick Sort** et le **Merge Sort**, sont moins sensibles à la structure initiale. Cependant, le **Quick Sort** peut connaître des dégradations importantes dans son pire cas, lorsque le pivot choisi divise mal les données (par exemple, si le tableau est déjà trié). Ce problème peut être atténué avec des optimisations telles que le choix du pivot médian.

### Taille des Données

La taille des tableaux influence considérablement le choix de l’algorithme. Sur des **petits tableaux**, les algorithmes simples, notamment l’**Insertion Sort**, peuvent être très compétitifs. Leur implémentation est légère, et ils s’exécutent rapidement pour de petites quantités de données, avec peu de surcharge liée aux appels récursifs ou à la gestion des structures intermédiaires.

En revanche, pour des **tableaux volumineux**, les algorithmes plus avancés, comme le **Quick Sort** et le **Merge Sort**, s’imposent. Leur complexité de $O(n \log n)$ leur permet de traiter efficacement de grandes quantités de données. Ces algorithmes sont également mieux adaptés aux scénarios nécessitant des performances constantes, même avec des données aléatoires ou mal organisées.

---

## Règles de Bonnes Pratiques

### Choix selon la taille

La taille des tableaux est un critère déterminant dans le choix de l’algorithme. Pour des **petits tableaux** ou des données **presque triées**, l’**Insertion Sort** est souvent le meilleur choix. Il offre une faible surcharge et peut être très rapide dans ces cas spécifiques. 

Pour des **tableaux généraux**, l’**Insertion Sort** devient inefficace, et le **Quick Sort** s’impose comme la meilleure option grâce à sa complexité moyenne en \(O(n \log n)\) et ses performances généralement excellentes. Si la **stabilité** est une contrainte (par exemple, pour préserver l’ordre des éléments égaux), il est recommandé d’utiliser le **Merge Sort**, qui garantit cette propriété.


### Gérer les cas extrêmes

Certains algorithmes, comme le **Quick Sort**, peuvent connaître des performances dégradées dans des cas particuliers, comme des tableaux déjà triés. Pour éviter ces situations, il est conseillé d’implémenter un **Quick Sort optimisé** avec une stratégie de pivot améliorée, telle que le choix du pivot médian ou un pivot aléatoire.

En revanche, pour des scénarios où des performances **constantes** sont essentielles, le **Heap Sort** est un choix sûr. Sa complexité en \(O(n \log n)\) est stable quelle que soit la structure des données.
