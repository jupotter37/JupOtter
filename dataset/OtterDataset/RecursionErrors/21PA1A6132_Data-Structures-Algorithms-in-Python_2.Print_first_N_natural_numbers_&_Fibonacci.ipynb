```python
def print_1_N(n):
  if n==0:
    return
  print_1_N(n-1)
  print(n,end=" ")

n = int(input())
print_1_N(n)
```

```python
def print_N_1(n):
  if n==0:
    return
  print(n,end=" ")
  print_N_1(n-1)

n = int(input())
print_N_1(n)
```

```python
def fib(n):
  if n==1 or n==2:
    return 1
  return fib(n-1)+fib(n-2)

n = int(input())
print(fib(n))
```

```python
def fact(n):
  if n==0 or n==1:
    return 1
  return n*fact(n-1)

print(fact(int(input())))
```
