## 재귀 함수

```python
def selfCall():
    print('하', end = '')
    selfCall()   # 중복되어 끝까지 호출한다.
selfCall()
```

```python
def count(num) :
    if num >= 1:
        print(num, end = ' ')
        count(num-1)
    else : 
        return
count(10)
count(20)
```

```python
def factorial(num) :
    if num <= 1:
        return num
    else :
        return num * factorial(num-1)
print(factorial(4))
print(factorial(10))
```

## 필터 함수

```python
def adult_func(n):
    if n>= 19:
        return True
    else :
        return False

ages = [34,39,20,18,13,54]
print('성년 리스트 : ')
for i in filter(adult_func, ages):
    print(i, end=' ')
```

```python
# lambda를 이용한 간결한 방법

ages = [34,39,20,18,13,54]

print('성년 리스트 : ')
for i in filter(lambda x:x >= 19, ages):
    print(i, end = ' ')
```

## reduce 함수

```python
def reduce(function, iterable, initializer = None):
    it = iter(iterable)
    if initializer is None :
        value = next(it)
    else :
        value = initializer
    for element in it :
        value = function(value, element)
    return value

def sum(x,y) :
    return x+y

reduce(sum, [1,2,3,4,5])
```
