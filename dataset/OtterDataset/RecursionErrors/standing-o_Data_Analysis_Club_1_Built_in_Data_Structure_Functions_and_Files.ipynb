## 3. Enumerate, Sorted, Zip

```python
tmp = ['a', 'b', 'c']
mapping = {}
for i, v in enumerate(tmp) : 
    mapping[v] = i
    print(mapping)
```

## 7. Function

```python
def func():
    a = []
    for i in range(5):
        a.append(i)
    
    print(a)
```

```python
func() 
```

- return multiple value : f(x,y)

```python
def my_func(x,y):
    x = x+1
    y = y+2
    return x, y
```

```python
a, b = my_func(1,1)
print(a, b)
```

```python
def def_sin(n=10):
    print("what is generator?")
    for i in range(1,n+1):
        yield i**2
```
