```python
turns = [
  RIGHT,
  UP,
  LEFT,
  DOWN,
]
turn_to_i = {j: i for i, j in enumerate(turns)}

def turn(dir, s):
  pos = turn_to_i[dir]
  if s == 'L':
    pos -= 1
  elif s == 'R':
    pos += 1
  else:
    raise Exception()
  return turns[(pos + 4) % 4]
```

```python
def go_N(p, dir, N):
  for i in range(N):
    p, dir = go(p, dir)
  return p, dir

def go(p, dir):
  pp = p + dir
  if pp in exist:
    return pp, dir
  
  # find fulcrum to left
  pp = p
  dir_to_left = turn(dir, 'L')
  X = None
  for i in range(100):
    x = pp.x % 50
    y = pp.y % 50
    if any([
      dir == RIGHT and y == 0,
      dir == UP and x == 0,
      dir == LEFT and y == 49,
      dir == DOWN and x == 49,
    ]):
    # if (x == 0 or x == 49) and (y == 0 or y == 49):
      X = i
      break
    pp, dir_to_left = go(pp, dir_to_left)
  X += 1
  print(f'found X {X}')
  
  dir = turn(dir, 'L')
  p, dir = go_N(p, dir, X)
  dir = turn(dir, 'R')
  p, dir = go_N(p, dir, X)
  dir = turn(dir, 'L')

  return p, dir

p = Point(99, 99)
dir = RIGHT

go(p, dir)
```
