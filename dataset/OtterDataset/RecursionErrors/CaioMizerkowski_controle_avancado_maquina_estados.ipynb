```python
import numpy as np

def gen_state(state,to,prob):
    def new_state():
        p = np.random.rand()
        print(f'{state}:{p}')
        if p < prob:
            return to
    return new_state

states = {
    'A': gen_state('A', 'B', 0.9),
    'B': gen_state('B', 'A', 0.9)
}

def rec(state):
    if state:
        return rec(states[state]())
    print('End')

rec('A')
```
