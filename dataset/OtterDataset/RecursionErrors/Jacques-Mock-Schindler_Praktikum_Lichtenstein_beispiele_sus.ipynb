## Beispiel THE SHINING

```python
def shining():
    print("All work and no play"
          "makes Jack a dull boy.")
    shining()
```

```python
shining()
```

## Beispiel THE SHINING (ungruselig)

```python
def shining_u(stop):
      if stop == 0: # Abbruchbedingung
            return
      
      print("All work and no play"
          "makes Jack a dull boy.")
      stop -= 1
      shining_u(stop)
```

```python
shining_u(5)
```

## Beispiel GAUSS

```python
def gauss(n):
    if n == 1:
        return 1
    else:
        return gauss(n-1) + n
```

```python
print(gauss(5))
```

## Beispiel FAKULTAET

```python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return factorial(n-1) * n
```

```python
print(factorial(5))
```
