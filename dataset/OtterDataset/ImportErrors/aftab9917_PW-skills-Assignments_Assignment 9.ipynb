```python
#ANSWER 3

'''ArithmeticError is a base class for numeric errors that might occur during arithmetic operations. Two common errors defined in this class are:

ZeroDivisionError: Raised when the second operand of a division or modulo operation is zero.'''


try:
    result = 5 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
#OverflowError: Raised when an arithmetic operation exceeds the limits of the data type.

try:
    result = 2 ** 1000  # OverflowError for exceeding integer limit
except OverflowError as e:
    print(f"Error: {e}")
```

```python
#ANSWER 4

#LookupError is a base class for errors that occur when a key or index is not found. Two common subclasses are:

#KeyError: Raised when a dictionary key is not found.


my_dict = {'a': 1, 'b': 2}
try:
    value = my_dict['c']
except KeyError as e:
    print(f"Error: {e}")
#IndexError: Raised when a sequence subscript is out of range.


my_list = [1, 2, 3]
try:
    value = my_list[4]
except IndexError as e:
    print(f"Error: {e}")
```

```python
#ANSWER 5

#ImportError is a base class for exceptions raised when an import statement fails. It has a subclass called ModuleNotFoundError, which is more specific and raised when a module is not found during the import.


try:
    import non_existent_module
except ImportError as e:
    print(f"ImportError: {e}")

try:
    from non_existent_module import some_function
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError: {e}")
#In the first example, an ImportError is caught when trying to import a non-existent module. In the second example, a more specific ModuleNotFoundError is caught when attempting to import a non-existent function from a module.
```
