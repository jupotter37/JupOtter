Write a function `find_trains` to find the trains from a source station to destination station. The function takes two arguments, the code of the source station and the code of the destination station and returns the dataframe containing all the trains from these two stations.

````python
def find_trains(from_station_code, destination_station_code):
    stations_matching_origin = df_trains[df_trains.from_station_code == from_station_code]

    # pick from the 'stations_matching_origin' the ones that match the destination_station_code
    matching_trains = stations_matching_origin[stations_matching_origin.to_station_code == destination_station_code]
    return matching_trains
````
