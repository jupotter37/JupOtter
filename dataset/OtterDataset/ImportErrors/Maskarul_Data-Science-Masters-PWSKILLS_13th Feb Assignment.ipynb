### Q6. List down some best practices for exception handling in python.

> 1. use try, except, else and finally block to catch and handle exceptions and do more things with code the exception handling scenario.
> 2. Catching specific exceptions is crucial for writing maintainable and easy-to-debug code. By specifying the exact exception types you expect to handle, you can avoid inadvertently suppressing unexpected exceptions that may reveal underlying issues in your code.
> 3. In some cases, you may want to create your own custom exceptions to handle specific error conditions that are unique to your application. Custom exceptions can make your code more readable and easier to understand by providing more descriptive error messages.
> 4. Avoid using overly broad exception handling, which occurs when you catch all exceptions using a generic except block without specifying the exact exception type. This can lead to unintentionally suppressing unexpected exceptions and make it difficult to identify and fix underlying issues.
