Q1. Explain why we have to use the Exception class while creating a Custom Exception.

We use the Exception class when creating a custom exception to ensure our custom exception inherits all the built-in functionalities of standard exceptions. This provides a consistent interface for error handling, making it easier to catch, raise, and manage exceptions in a uniform way within the program.

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
#Two common exceptions derived from ArithmeticError are ZeroDivisionError and OverflowError.
#1.ZeroDivisionError: This error occurs when a division or modulo operation is attempted with a divisor of zero.
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e) 
#2.OverflowError: This error occurs when the result of an arithmetic operation exceeds the limits of the numeric type.
import math
try:
    result = math.exp(1000)
except OverflowError as e:
    print("Error:", e)  
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

#The LookupError class is a built-in exception in Python that serves as the base class for errors
#raised when a lookup operation fails, such as looking up a key in a dictionary or an index in a list.

#KeyError: Raised when trying to access a dictionary with a key that doesn't exist.
my_dict = {'a': 1, 'b': 2}
try:
    value = my_dict['c']
except KeyError as e:
    print("Error:", e)  # Output: Error: 'c'
    
#IndexError: Raised when trying to access a list element with an index that is out of range.
my_list = [1, 2, 3]
try:
    element = my_list[5]
except IndexError as e:
    print("Error:", e)  # Output: Error: list index out of range
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a built-in exception in Python raised when an import statement fails to find the module definition or when a from ... import ... fails to find a name that is being imported.

ModuleNotFoundError is a subclass of ImportError that specifically occurs when the module being imported cannot be found. It was introduced in Python 3.6 to provide a more specific exception for this type of import failure.
