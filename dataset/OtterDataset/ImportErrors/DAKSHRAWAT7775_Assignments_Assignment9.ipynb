```python
'''3. The ArithmeticError class in Python is the base class for exceptions that arise during arithmetic operations. It serves as a general   
category for arithmetic-related errors. Two specific errors that are derived from ArithmeticError are OverflowError and ZeroDivisionError.
These errors, derived from the ArithmeticError class, help to categorize and handle specific issues related to arithmetic operations. It's   
important to catch and handle these exceptions appropriately to prevent unexpected program termination and to provide meaningful feedback or  
take corrective actions.'''
try:
    result = 10 / 0 
except ZeroDivisionError as zde:
    print("Caught a ZeroDivisionError")
import math
try:
    result = math.exp(1000)  
except OverflowError as oe:
    print("Caught an OverflowError")
```

```python
'''4. 
The LookupError class in Python is a base class for exceptions that arise when a key or index is not found during a lookup operation.
LookupError itself is not meant to be instantiated directly but serves as a base class for more specific lookup-related exceptions, such as KeyError and IndexError.'''
my_dict = {'key1': 'value1', 'key2': 'value2'}
try:
    value = my_dict['key3'] 
except KeyError as ke:
    print("Caught a KeyError")
my_list = [1, 2, 3]
try:
    value = my_list[4]  
except IndexError as ie:
    print(f"Caught an IndexError")
```
