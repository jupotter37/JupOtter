```python
def read_file(afile):
    df = pd.read_csv(afile, sep='\t')
    return df


def read_binding_db(file_list):
    with mp.Pool(64) as pool:
    
        df_list = list(tqdm(pool.imap_unordered(read_file, file_list), total=len(file_list)))

    df = pd.concat(df_list)
#     print(df.head())
#     df = df.sample(frac=0.0001)
    
    return df


def convert_fp_str(fp_str):
    
    try:
        rdkit_fp = CreateFromBinaryText(fp_str)
        fp = np.frombuffer(rdkit_fp.ToBitString().encode(), 'u1') - ord('0')
        return fp
    except Exception as e:
        print(e)
        return

def np_to_bv(fv):
    bv = DataStructs.ExplicitBitVect(len(fv))
    for i,v in enumerate(fv):
        if v:
            bv.SetBit(i)
    return bv
    
def tanimoto_dist(fp1, fp2):    
    if isinstance(fp1, np.ndarray):
        fp1 = np_to_bv(fp1)
    if isinstance(fp2, np.ndarray):
        fp2 = np_to_bv(fp2)
    
    return 1 - FingerprintSimilarity(fp1,fp2)


def compute_HD_encoding(fp, D=1000, Q=5, batch_size=1):
    
    if isinstance(fp, str):
        fp = np.frombuffer(fp.ToBitString().encode(), 'u1') - ord('0')
    fp = fp.reshape(1,-1)
    fp, _ = tfHD.normalize(fp)
    return encode_utils.encode_single(fp, D=D, Q=Q, batch_size=batch_size, tf_device_str="CPU", return_time=False).numpy()



def compute_HD_encoding_batch(fp_array, D=1000, Q=5, batch_size=1):
    fp_array, _ = tfHD.normalize(fp_array)
    return encode_utils.encode_batch(fp_array, D=D, Q=Q, batch_size=batch_size, tf_device_str="CPU", return_time=False).numpy()


def compute_HD_dist(fp1, fp2):
    hd_fp1 = compute_HD_encoding(fp1)
    hd_fp2 = compute_HD_encoding(fp2)
    
#     dist = np.linalg.norm(hd_fp1 - hd_fp2, ord=2)
    dist = distance.cosine(hd_fp1, hd_fp2)
    return dist


def compute_pairwise_dists(fp_tup):
    fp1, fp2 = fp_tup
    
    if isinstance(fp1, rdkit.DataStructs.cDataStructs.ExplicitBitVect):
        fp1 = np.frombuffer(fp1.ToBitString().encode(), 'u1') - ord('0')
    
    if isinstance(fp2, rdkit.DataStructs.cDataStructs.ExplicitBitVect):
        fp2 = np.frombuffer(fp2.ToBitString().encode(), 'u1') - ord('0')
    
    hd_dist = compute_HD_dist(fp1, fp2)
    tani_dist = tanimoto_dist(fp1, fp2)
    
    return hd_dist, tani_dist


def make_scatter_dist_plot(dist_array):
    f, ax = plt.subplots(1,2, figsize=(12,10))
    ax=ax.flatten()
    sns.scatterplot(x=dist_array[:, 0], y=dist_array[:,1], ax=ax[0], label=pearsonr(dist_array[:, 0], 
                                                                                dist_array[:,1]))
    sns.distplot(x=dist_array[:, 0], ax=ax[1])
    plt.legend()
    
    return f, ax


    
def convert_logki_kd_to_class(y):
    if y > 8:
        return 1
    elif y < 6:
        return 0
    else:
        return 2
 
    
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
def compute_pca(X):
    
    pca_pipe = Pipeline([("std_scaler_0", StandardScaler()), ("pca_0", PCA(n_components=2))])
    comps = pca_pipe.fit_transform(X)
    return comps
    
```

```python
def read_binding_db_ligand_fp():
    
    p = Path("/g/g13/jones289/workspace/hd-cuda-master/datasets/bindingDB/MolSMSfingerprint")
    file_list = list(p.glob("*.tsv"))
    
    print(f"{len(file_list)} files to process for bindingDB..")
    bind_db_df = read_binding_db(file_list)
    
    print(bind_db_df.shape)
    
    with mp.Pool(64) as pool:
        bind_db_fp_list = list(tqdm(pool.imap(convert_fp_str, bind_db_df['fingerprint(1024)'].values.tolist()), total=len(bind_db_df)))

#     bind_db_fp_list = [x for x in bind_db_list if x is not None]
    
    bind_db_df['np_fp'] = bind_db_fp_list
    # drop null values
    bind_db_df = bind_db_df.drop(bind_db_df[bind_db_df['np_fp'].isnull()].index)
    
    return bind_db_df
    
    

bind_db_full_fp_df = read_binding_db_ligand_fp()
```

```python
def bind_db_main(verbose=True, nrows=None, m_type=None):
    
    assert m_type in ['Ki (nM)', 'IC50 (nM)', 'EC50 (nM)']
    
    # /usr/WS1/jones289/hd-cuda-master/datasets/bindingDB/BindingDB_All.tsv
    # 'Ligand SMILES', 'Ligand InChI', 'Ligand InChI Key', 'BindingDB MonomerID', 'Number of Protein Chains in Target (>1 implies a multichain complex)'
    # 'Ki (nM)','IC50 (nM)', 'Kd (nM)', 'EC50 (nM)', 'kon (M-1-s-1)', 'koff (s-1)'

    df = pd.read_csv("/usr/WS1/jones289/hd-cuda-master/datasets/bindingDB/BindingDB_All.tsv", nrows=10, 
                    sep='\t', engine='c')
    
    if verbose:
        for column in df.columns:
            print(column)
    
    df = pd.read_csv("/usr/WS1/jones289/hd-cuda-master/datasets/bindingDB/BindingDB_All.tsv", nrows=nrows,
                     sep='\t', usecols=['BindingDB Reactant_set_id', 'Ligand SMILES', 'Ligand InChI Key','BindingDB MonomerID',
                                        'Ki (nM)', 'IC50 (nM)', 'EC50 (nM)',
                                        'Link to Target in BindingDB',
                                       'BindingDB Target Chain  Sequence'], engine='c')
    
    print(f"{df['BindingDB Target Chain  Sequence'].unique().shape} protein sequences detected")    
    
    # drop rows without the 'Ki (nM)', 'IC50 (nM)', 'EC50 (nM)' binding data
    df = df.drop(df[df[['Ki (nM)', 'IC50 (nM)', 'EC50 (nM)']].isnull().values.sum(axis=1) == 3].index)
    df = df.drop(df[df[m_type].isnull()].index)
    
    
    # drop rows that have bounded measurements, i.e. '<', '>' or other non-numeric values
    df[m_type] = pd.to_numeric(df[m_type], errors='coerce')
    df[f"-log({m_type})"] = df[m_type].apply(lambda x: -1 * np.log10(x))
    
    
    df = pd.merge(df, bind_db_full_fp_df, left_on='Ligand SMILES', right_on='SMILES')
    
    df = df[~df[f"-log({m_type})"].isin([np.nan, np.inf, -np.inf])]
    
    #remove large values
    df = df.drop(df[df[m_type].apply(lambda x: x >0.2)].index)

#     df = df.replace([np.inf, -np.inf], np.nan, inplace=True)

#     df = df.replace([np.inf, -np.inf], np.nan).dropna(subset=[f"-log({m_type})"], how="all")
    
    
    f, ax = plt.subplots(2,1, figsize=(12,6))
    ax = ax.flatten()
    sns.distplot(df[m_type], ax=ax[0])
    sns.violinplot(df[f"-log({m_type})"], ax=ax[1])
    
    
#     with mp.Pool(64) as pool:
#         bind_db_list = list(tqdm(pool.imap(convert_fp_str, bind_db_df['fingerprint(1024)'].values.tolist()), total=len(bind_db_df)))

#     bind_db_list = [x for x in bind_db_list if x is not None]
    
#     combo_list = list(itertools.product(bind_db_list, bind_db_list))
    
#     dist_result_list = []
#     for fp_tup in tqdm(combo_list[:100]):
#         dist_result_list.append(compute_pairwise_dists(fp_tup))
        
#     dist_array = np.concatenate(dist_result_list).reshape(len(dist_result_list), -1)
        
#     make_scatter_dist_plot(dist_array)
    return df
    print("done.")
```
