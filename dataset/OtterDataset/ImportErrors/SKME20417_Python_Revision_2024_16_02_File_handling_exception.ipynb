```python
import sys
lst = [2,0,'hello',3 ,None]


for i in lst:
    try:
        print("Current element :", i)
        res = 5/int(i)
        print("result :", res)
    except ValueError as ve:
        print('this is value error')
        
    except ZeroDivisionError as ze:
        print("Hey you can not divide element by zero")
        
    except Exception as e:
        print("error occured :", e)
        print(sys.exc_info()[0])
    finally:
        #db.close
        #api.close
        #file.close
        print("I will always be execueted..")
    print("="*50)
    print()
    
print("100 lines of code..")
```
