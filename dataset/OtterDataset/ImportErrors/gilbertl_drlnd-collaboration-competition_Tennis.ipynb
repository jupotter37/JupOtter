````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents 
num_agents = len(env_info.agents)
print('Number of agents:', num_agents)

# size of each action
action_size = brain.vector_action_space_size
print('Size of each action:', action_size)

# examine the state space 
states = env_info.vector_observations
state_size = states.shape[1]
print('There are {} agents. Each observes a state with length: {}'.format(states.shape[0], state_size))
print('The state for the first agent looks like:', states[0])
````

### 3. Train the Agent

````python
from ddpg_agent import Agent
from collections import deque, Counter
import torch
import matplotlib.pyplot as plt
%matplotlib inline
from workspace_utils import active_session
import datetime
from decimal import Decimal
import uuid


# agent = Agent(actor_state_size=state_size, action_size=action_size, random_seed=2,
#               actor_fc1_units=100, actor_fc2_units=100, critic_fc1_units=200, critic_fc2_units=200,
#               lr_actor=11e-5, lr_critic=11e-4, buffer_size=int(1e7), tau=2e-2)

# Resume from previous training session.
# agent.actor_local.load_state_dict(torch.load('checkpoint_actor_continue.pth'))
# agent.actor_target.load_state_dict(torch.load('checkpoint_actor_continue.pth'))
# agent.critic_local.load_state_dict(torch.load('checkpoint_critic_continue.pth'))
# agent.critic_target.load_state_dict(torch.load('checkpoint_critic_continue.pth'))


def ddpg(agent, agent_id, n_episodes=3500, print_every=50):
    scores_deque = deque(maxlen=print_every)
    scores = []
    for i_episode in range(1, n_episodes+1):
        env_info = env.reset(train_mode=True)[brain_name] 
        initial_states = env_info.vector_observations  # 2x24
        actor_state_0 = initial_states[0].reshape(1, -1)
        actor_state_1 = initial_states[1].reshape(1, -1)
        agent.reset()
        both_scores = [0, 0]
        done = False
        while not done:
            action_0 = agent.act(actor_state_0)
            action_1 = agent.act(actor_state_1)
            both_actions = np.vstack((action_0, action_1))
            env_info = env.step(both_actions)[brain_name]
            next_states = env_info.vector_observations
            actor_next_state_0 = next_states[0].reshape(1, -1)
            actor_next_state_1 = next_states[1].reshape(1, -1)
            rewards = env_info.rewards
            reward_0 = rewards[0]
            reward_1 = rewards[1]
            done = np.any(env_info.local_done)
            agent.step(
                actor_state_0, actor_state_1, action_0, reward_0,
                actor_next_state_0, actor_next_state_1, done)
            actor_loss, critic_loss = agent.step(
                actor_state_1, actor_state_0, action_1, reward_1, 
                actor_next_state_1, actor_next_state_0, done)
            actor_state_0 = actor_next_state_0
            actor_state_1 = actor_next_state_1
            both_scores = np.sum([both_scores, rewards], axis=0)
        better_score = np.max(both_scores)
        scores_deque.append(both_scores)
        scores.append(both_scores)
        average_scores = np.mean(scores_deque, axis=0)
        print('\rEpisode {}\tAverage Score: [{:.2f}, {:.2f}]'.format(i_episode, average_scores[0], average_scores[1]), end="")
        torch.save(agent.actor_local.state_dict(), 'checkpoint_actor_{}.pth'.format(agent_id))
        torch.save(agent.critic_local.state_dict(), 'checkpoint_critic_{}.pth'.format(agent_id))
        if i_episode % print_every == 0:
            print('\rEpisode {}\tAverage Score Total: {:.2f}'.format(
                i_episode, np.mean(np.max(scores_deque, axis=1))))
            print('Episode {}\tActor Loss: {:.2f}\tCritic Loss: {:.2f}'.format(
                i_episode, actor_loss, critic_loss))
            print(datetime.datetime.now())
    
    return np.max(scores, axis=1)

with active_session():
    while True:
        agent_id = uuid.uuid4()
        lr_actor = np.random.uniform(5e-5, 12e-5)
        lr_critic = np.random.uniform(5e-4, 1e-3)
        tau = np.random.uniform(5e-3, 7e-3)
        agent = Agent(actor_state_size=state_size, action_size=action_size, random_seed=2,
              actor_fc1_units=100, actor_fc2_units=100, critic_fc1_units=200, critic_fc2_units=200,
              lr_actor=lr_actor, lr_critic=lr_critic, buffer_size=int(1e7), tau=tau)

        print("Agent {} Using parameters lr_actor {:.2e}, lr_critic {:.2e}, tau {:.2e}".format(
            agent_id, Decimal(lr_actor), Decimal(lr_critic), Decimal(tau)))

        print("Start: {}".format(datetime.datetime.now()))
        scores = ddpg(agent, agent_id)

        fig = plt.figure()
        ax = fig.add_subplot(111)
        plt.plot(np.arange(1, len(scores)+1), scores)
        plt.ylabel('Score')
        plt.xlabel('Episode #')
        plt.show()
        print("End: {}".format(datetime.datetime.now()))
        print("\n")
````
