```python
import requests
from bs4 import BeautifulSoup
import json
import regex as re

url = "https://www.ebay.com/globaldeals"

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
product_elements = soup.find_all('div', class_='dne-itemtile-detail', limit=5)

products = []

for product in product_elements:
    title = product.find('h3', class_='dne-itemtile-title ellipse-2')
    price = product.find('span', itemprop='price').text.strip()
    products.append({"title": title, "price": price})

products.sort(key=lambda x: float(x['price'].replace('US $', '').replace(',', '')))

for product in products:
    print(product)

with open('sorted_products.json', 'w') as json_file:
    json.dump(products, json_file, indent=4)

```
