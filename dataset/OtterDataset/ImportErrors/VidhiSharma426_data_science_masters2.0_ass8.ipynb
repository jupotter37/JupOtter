Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print("Python Exception Hierarchy:")
print_exception_hierarchy(BaseException)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

```python
# the arithmetic errors class control numeric errors
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:", e)  # Output: Error: division by zero

    
    
import math

try:
    result = math.exp(1000)
except OverflowError as e:
    print("Error:", e)  # Output: Error: math range error
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# LookupError is the base class for exceptions that occur when a mapping or sequence key is not found in a collection.
# KEY ERROR
my_dict = {1:"vidhi",2:"arpita"}
try:
    my_dict[3]
except Exception as e:
    print(e)
# INDEX ERROR
list1=[1,2,3,4]
try:
    list1[5]
except Exception as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
# In Python, ImportError is an exception that is raised when an import statement fails to import a module or a name from a module
# and the module not found error is the subclass of import error
try:
    import class1
except ImportError as e:
    print(e)
    
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print("Module Not Found Error:", e) 
```
