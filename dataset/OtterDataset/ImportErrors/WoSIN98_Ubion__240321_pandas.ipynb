### 데이터프레임의 필터링 
- loc [ 인덱스 조건식, 컬럼 조건식 ]
    - 인덱스의 값이나 컬럼의 값을 기준으로 필터링 (df.loc[1, 'A'])
    - 인덱스 조건식이나 컬럼 조건식을 기준으로 필터링 
- iloc [ 인덱스 위치, 컬럼 위치 ]
    - 인덱스 위치, 컬럼 위치를 기준으로 필터링 (df.iloc[2:4, 0:2])
