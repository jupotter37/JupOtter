````python
import sqlite3

# Connect to SQLite database. It will be created if it doesn't exist.
conn = sqlite3.connect('vehicles.db')

# Create a cursor object
c = conn.cursor()

# Create table
c.execute('''
    CREATE TABLE vehicles (
        vehicle_id TEXT,
        make TEXT,
        model TEXT,
        type TEXT,
        vin TEXT,
        year INT,
        current_mileage INT,
        current_hours INT
    );
''')

# Insert data
c.execute("INSERT INTO vehicles VALUES ('JD6220', 'John Deere', '6220', 'Tractor', 'AB12', 2000, NULL, 5100)")
c.execute("INSERT INTO vehicles VALUES ('CAT906M', 'Cat', '906M', 'Loader', '5', 2021, 3200, 3000)")

# Commit the changes
conn.commit()

# Close the connection
conn.close()
````

````python
# Create a DataFrame for vehicles
vehicles_data = {
    'vehicle_id': ['JD6220', 'CAT906M'],
    'make': ['John Deere', 'Cat'],
    'model': ['6220', '906M'],
    'type': ['Tractor', 'Loader'],
    'vin': ['AB12', '5'],
    'year': [2000, 2021],
    'current_mileage': [None, 3200],
    'current_hours': [5100, 3000]
}

vehicles_df = pd.DataFrame(vehicles_data)

# Create a DataFrame for services
services_data = {
    'service_id': ['Oil_Change_JD6220'],
    'vehicle_id': ['JD6220'],
    'service_name': ['Oil_Change'],
    'months_between_service': [12],
    'miles_between_service': [None],
    'hours_between_service': [500]
}

services_df = pd.DataFrame(services_data)

# Create a DataFrame for parts
parts_data = {
    'service_id': ['Oil_Change_JD6220', 'Oil_Change_JD6220'],
    'vehicle_id': ['JD6220', 'JD6220'],
    'part_number': ['RE504836', '12345678'],
    'quantity': ['1', '4.2 Gallons (16L)'],
    'part_description': ['Oil Filter', 'Plus-50'],
    'part_hyperlink': ['https://www.google.com/', 'https://www.google.com/']
}

parts_df = pd.DataFrame(parts_data)

# Create a DataFrame for service_records
service_records_data = {
    'service_id': ['Oil_Change_JD6220', 'Oil_Change_JD6220'],
    'vehicle_id': ['JD6220', 'JD6220'],
    'service_date': ['2024-02-01', '2023-08-18'],
    'service_mileage': [None, None],
    'service_hours': [4500, 4100]
}

service_records_df = pd.DataFrame(service_records_data)
````

````python
merged_df = pd.merge(vehicles_df, services_df, on='vehicle_id')
merged_df = pd.merge(merged_df, parts_df, on='service_id')
merged_df = pd.merge(merged_df, service_records_df, on='service_id')

# Find the maximum service hours for each vehicle
max_service_hours = merged_df.groupby('vehicle_id')['service_hours'].max()
max_service_hours = max_service_hours.values[0]
````

````python
# Find the vehicles that are due for service
due_for_service_vehicles = vehicles_df[vehicles_df['current_hours'] > max_service_hours]

print(due_for_service_vehicles)
````
