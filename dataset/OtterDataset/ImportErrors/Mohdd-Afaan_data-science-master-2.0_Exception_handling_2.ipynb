Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans: LookupError exception forms the base class for all the exception that are raised due to KeyError or IndexError.

```python
# IndexError
try:
  list1 = [1, 2, 3]
  value = list1[5]
except LookupError as e:
  print(f'LookupError:{e}')
```

```python
# KeyError
try:
  d1 = {"name":"afaan","enrolled_in":"DSM2.0"}
  d1["age"]
except LookupError as e:
  print(f"LookupError: {e}")
```
