[voted-kaggle-dataset](https://www.kaggle.com/canggih/voted-kaggle-dataset/version/2#voted-kaggle-dataset.csv)

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import gensim
```

```python
df = pd.read_csv("voted-kaggle-dataset.csv")
df.head(5)
```

```python
print(f"length of texts = {len(df)}")
print(df.loc[10, "Description"])
```

```python
topics_data = df["Description"].copy()
topics_data.dropna(inplace=True)
```

```python
three_words_pattern = r"\b\w{3,}\b"
vectorizer = CountVectorizer(
    min_df=20, 
    stop_words="english", 
    token_pattern=three_words_pattern
)
vectorizer.fit(topics_data)
```

```python
topics_data_vectorized = vectorizer.transform(topics_data)
print(topics_data_vectorized)
```

```python
corpus = gensim.matutils.Sparse2Corpus(
    topics_data_vectorized, 
    documents_columns=False)
```

```python
id_to_word = dict((i, j) for j, i in vectorizer.vocabulary_.items())
id_to_word
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(
    corpus, 
    num_topics=6, 
    id2word=id_to_word, 
    passes=25, 
    random_state=0
    )
```

```python
ldamodel.print_topics(
    num_topics=6, 
    num_words=10
    )
```

```python
topics_names = ['Data Trends',
                'Sports Analytics',
                'High Education',
                'Data Analysis',
                'Data Overview',
                'CSV Dataset Analysis']
```

```python
new_doc = ["\n In the dynamic landscape of higher education, universities and \n\
colleges play a pivotal role in shaping the intellectual landscape.\n \
Whether it's a prestigious university, a state institution, or a specialized college,\n\
each serves as a bastion of knowledge, fostering innovation and contributing to the rich\n\
tapestry of academia. From the technological advancements in institutes to the academic rigor in California,\n\
these diverse educational entities collectively propel us into a future where data-driven insights \n\
and interdisciplinary approaches take center stage."]

print(new_doc[0])
```

```python
test_data_vectorized = vectorizer.transform(new_doc)

new_doc_corpus = gensim.matutils.Sparse2Corpus(
    test_data_vectorized,
    documents_columns=False)
doc_topics = ldamodel.get_document_topics(new_doc_corpus)
list(doc_topics)
```

```python
import numpy as np


def elicit_topic_name(doc_topics):
    return topics_names[np.squeeze(np.array(doc_topics))[:, 1].argmax()]


elicit_topic_name(doc_topics)
```
