We discussed the definition of hypothesis testing as "an idea that can be tested."

Criteria for a Good Null Hypothesis:

A good null hypothesis should be:
Clear and Specific: Clearly states what is being tested (e.g., "There is no difference in test scores...").
Testable: Involves measurable variables that can be collected and analyzed.
Falsifiable: Can be rejected if data shows otherwise.
Difference Between Null and Alternative Hypotheses:

Null Hypothesis (H): Assumes no effect, no difference, or no relationship between variables (e.g., "There is no difference in test scores...").
Alternative Hypothesis (H or H): Proposes that there is an effect, difference, or relationship, and is what you are trying to prove with data (e.g., "There is a difference in test scores...").

- Population: A group of people or a set you are interested in studying. For example if you wanted to know how tall 12-year olds in Toronto are, then all the 12-year olds in Toronto would makeup my entire population.

- Parameter: A number describing something about the entire POPULATION, for eg if average height of all 12-year olds is 150cm, then 150cm would be my parameter for the population.
- Statistic: A number descirbing something about the SAMPLE, if the average height of 30 12-year olds you measured is 135cm, then the statistic would be 135 cm.

"Imagine a world where the null hypothesis is true" when calculating a p-value:

When calculating a p-value we are basically trying to find out if our estimate or guess of something being true or not is accurate or how accurate it is. Inorder to do this we start by creating a null hypothesis something that says that there is no difference, no effect between variables. For example if there is a new study method that helps students get better scores, and we want to test if thats actually true or not. Our null hypothesis would be "the new method does not help students get better scores, the scores remain the same as they did with previous study methods". Now inorder to comprehend if our Null hypothesis is wrong or not we first need to assume the null hypothesis is true. So we pretend that the new study method really doesn't affect the test scores of students. Now inorder to prove our null hypotheses right, we need to gather some evidence some data so we can take two groups of students one that utilize that new study method and one that utilizes previous study methods, and then compare there scores. Now here is where we calculate our p-value, the p-value tells us if the students did better utilizing the new study method, so got a better score, or it had no affect at all. If our null hypothesis is true that indiciates that the new study method really does not have any effect on students test scores. If the p-value is small that rejects our null hypothesis, indicating that the new study method does have an effect on test scores, giving better grades to students. Conversly, if our p-value is large this means that our assumption or null hypotheses was accurate.

#### Fisher's Tea Experiment


##### PROBLEM INTRODUCTION

Relationship between this experiment and the original with Fisher and Bristol
- In this experiment, we will repeat the famous statistical lore of statistician Ronald Fisher and Dr.Muriels story which involves cups of tea with milk. We will be testing whether or not STA130 students are able to identify whether or not STA130 students are correctly able to identify the sequence in which the tea was made, that is if the milk was poured first or the tea. Statistician Bristol proved she could tell the difference in taste, even though Fisher thought there was no difference in taste based on the order of pouring. The purpose of our study is to determine whether STA130 students' accuracy in identifying the pouring order is the result of pure chance or whether statistical evidence points to a higher guessing ability.

- Dr. Bristol was given eight cups of tea in Fisher's original experiment, and her ability to determine the sequence in which they were to be poured depended on perception as well as individual taste. With a bigger sample of 80 STA130 students, our experiment, in contrast, focuses on statistical analysis to determine the percentage of students that accurately recognize the pouring order. The essential question of whether pouring order affects taste perception is still present even in this different context.


Statements of the Null Hypothesis and Alternative hypothesis
- Null Hypotheses: H: STA130 students can not predict the sequence of what was poured first from the tea and milk. The proportion of students who can correctly identify the pouring order is equal to chance (p = 0.5).


- Alternative Hypotheses: H: STA130 students can predict the sequence of what was poured first from tea and milk. The proportion of students who can correctly identify the pouring order is greater than chance (p > 0.5).



##### QUANTITATIVE ANALYSIS

Methodology Code and Explanations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Given data
n_students = 80  # Total students
n_correct = 49   # Students who correctly identified the pour order
p_hat = n_correct / n_students  # Sample proportion

# Null hypothesis: p = 0.5 (no preference)
p_null = 0.5

# Number of simulations
n_simulations = 10000
simulated_correct_counts = np.random.binomial(n_students, p_null, n_simulations)

# Calculate the p-value
# P-value is the proportion of simulated correct counts that are as extreme or more extreme than the observed count
p_value = np.mean(simulated_correct_counts >= n_correct)

# Plotting the distribution of simulated correct counts
plt.figure(figsize=(10, 6))
sns.histplot(simulated_correct_counts, bins=30, kde=True, stat='density', color='skyblue')
plt.axvline(n_correct, color='red', linestyle='--', label='Observed Correct Counts (49)')
plt.title('Simulated Sampling Distribution of Correct Identifications')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()

# Display the results
print(f"Observed Proportion (p_hat): {p_hat:.4f}")
print(f"Simulated Proportion under Null Hypothesis: {p_null:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the p-value
if p_value < 0.05:
    conclusion = "There is significant evidence to suggest that Bristol's ability to identify the pour order is not due to chance."
else:
    conclusion = "There is not enough evidence to suggest that Bristol's ability to identify the pour order is different from what would be expected by chance."

print(f"Conclusion: {conclusion}")
```

Explanations

- The above code creates a histogram that shows the distribution of sample proportions under the null hypothesis. The plot that the code produces is a visual representation of the simulated distribution of correctly identified samples under the null hypothesis. The red dashed line displays the observed count of correct identifications (49). It is simpler to understand how the observed result fits into the anticipated distribution of results with the help of this graphic.

- The observed proportion of students (p_hat) resulted to be 0.6125 indicating that approximately 61.25% of students correctly identified the sequence of the pouring. 

- The stimulations resulted in a P-value of 0.0294. Due to such a low p-value, We reject the null hypothesis because the p-value is much smaller than the typical significance level of 0.05. 0.05 > 0.0294 This implies that the pouring order identification skills of STA130 students are statistically significant and not they are not just a result of chance.






##### FINDINGS AND DISCUSSION

Conclusion
- In conclusion, based on the results of the hypothesis test, there is compelling evidence that STA130 students are able to discern between the pouring order, and are more likely to actually identify the order of the sequence of pouring due to taste as oppose to just randomly guessing it. This result is consistent with Fisher's original investigation into the complexity of taste perception, showing that the ability to identify differences in taste is still significant even in a larger and more abstract population. If the p-value was greater than the typical significance level of 0.05 we could have had evidence that supported our null hypotheses and Fishers assumptions, however, even in a larger sample the results disproved the hypotheses.
