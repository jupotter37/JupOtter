A classification decision tree is used for classification tasks, where the goal is to assign input data to one of several discrete categories. On the other hand, linear regression is used for when the goal is to predict a continuous numerical value that indicates a meaningful quantity.

| **Aspect**           | **Classification Decision Tree**                  | **Multiple Linear Regression**                   |
|-----------------------|---------------------------------------------------|-------------------------------------------------|
| **Output**           | Categorical (e.g., class labels like "yes"/"no"). | Continuous numerical values (e.g., 50.5).      |
| **Use Case**         | Classification tasks.                             | Regression tasks.                               |
| **Decision Process** | Rule-based, hierarchical splits in the feature space. | Mathematical model based on linear relationships. |
| **Interpretation**   | Easy to understand decision paths (e.g., a flowchart). | Requires interpretation of coefficients.        |
| **Predictions Matter** | Class label or probabilities indicate **category**. | Output value represents a meaningful **magnitude**. |

In Decision Trees, these interactions happen naturally because splits are sequential and conditional:
- A feature's importance or role in prediction is affected by earlier splits.
- The tree does not need explicit interaction terms; instead, the tree structure itself models interactions.

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

| **Metric**     | **Focus**                 | **Best for Avoiding**    | **Example Context**          |
|-----------------|---------------------------|---------------------------|-------------------------------|
| **Accuracy**    | Overall correctness       | Balanced errors           | Common, low-risk conditions. |
| **Sensitivity** | Detecting all positives   | False negatives           | Critical early diagnosis.     |
| **Specificity** | Detecting all negatives   | False positives           | Pre-invasive procedures.      |
| **Precision**   | Correct flagged positives | False positives           | Spam detection, alerts.       |

Accuracy measures the proportion of all correct predictions (true positives + true negatives) out of all predictions. It gives a general idea of how well the model performs across all outcomes. This makes accuracy a balanced metric, as it accounts for both errors (false positives and false negatives) equally. For example accuracy can be applied to predicting the type of a pokemon from its base stats, where stakes are low.

In most situations, true positives and true negatives are the most desirable outcomes. However, there are scenarios where we must identify whichever is preferable, a false positive or a false negative.

In critical early diagnosis (sensitivity-focused), missing a disease (false negative) is more harmful than wrongly diagnosing someone (false positive). However, in pre-invasive procedures (specificity-focused), minimizing unnecessary procedures (false positives) might take precedence.

Precision evaluates how many of the flagged (or predicted) positives are actually correct. It emphasizes being right when the model says "positive." Precision is crucial when false positives are costly or disruptive, such as spam detection, fraud detection, or alert systems. If a fraud detection system falsely flags legitimate transactions, it can be very fustrating for users and cause operational bottlenecks.

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Define the predictor variables for both models using the test set
X_test_clf = ab_reduced_noNaN_test[['List Price']]
y_test_clf = ab_reduced_noNaN_test['Hard_or_Paper']

X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test_clf2 = ab_reduced_noNaN_test['Hard_or_Paper']

# Generate predictions
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices
cm_clf = confusion_matrix(y_test_clf, y_pred_clf)
cm_clf2 = confusion_matrix(y_test_clf2, y_pred_clf2)

# Sensitivity, Specificity, and Accuracy functions
def calculate_metrics(cm):
    tn, fp, fn, tp = cm.ravel()  # True Negative, False Positive, False Negative, True Positive
    
    # Sensitivity (Recall)
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    
    # Specificity
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    
    # Accuracy
    accuracy = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) != 0 else 0
    
    return sensitivity, specificity, accuracy

# Calculate metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(cm_clf)

# Calculate metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(cm_clf2)

# Print Confusion Matrices and Metrics
print("Confusion Matrix for clf (using List Price):")
print(cm_clf)
print(f"Sensitivity: {sensitivity_clf:.4f}, Specificity: {specificity_clf:.4f}, Accuracy: {accuracy_clf:.4f}")

print("\nConfusion Matrix for clf2 (using NumPages, Thick, List Price):")
print(cm_clf2)
print(f"Sensitivity: {sensitivity_clf2:.4f}, Specificity: {specificity_clf2:.4f}, Accuracy: {accuracy_clf2:.4f}")
````

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>

In a classification decision tree, the model builds by considering all possible decision rules at each step and selecting the one that best improves prediction accuracy, with the importance of each predictor variable measured by how much it contributes to the improvement. The .feature_importances_ attribute quantifies these contributions, allowing us to understand the relative significance of each feature, while methods like GridSearchCV help manage the risk of overfitting due to the model's complexity.

````python
import matplotlib.pyplot as plt
import numpy as np

# Get the feature names and their corresponding importances
features = clf2.feature_names_in_
importances = clf2.feature_importances_

# Sort the feature importances in descending order and get the sorted indices
indices = np.argsort(importances)[::-1]

# Plot the feature importances
plt.figure(figsize=(8, 6))
plt.bar(range(len(features)), importances[indices], align="center")
plt.xticks(range(len(features)), np.array(features)[indices], rotation=45)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importances in DecisionTreeClassifier (clf2)')
plt.show()

# Report the most important feature
most_important_feature = features[indices[0]]
print(f"The most important feature for predictions in clf2 is: {most_important_feature}")
````
