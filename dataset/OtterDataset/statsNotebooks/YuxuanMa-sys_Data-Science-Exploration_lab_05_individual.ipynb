## <u>Case Study 1</u>: Learning More about Recent Netflix Shows and Movies

The data contained in the NetflixOriginals.csv file consists of all Netflix original films released as of June 1st, 2021. Additionally, it also includes all Netflix documentaries and specials. The IMDB scores of each of these films and shows is also given.

The dataset is comprised of the following variables.
* Title of the film
* Genre of the film
* Original premiere date
* Runtime in minutes
* IMDB scores (as of 06/01/21)
* Languages currently available (as of 06/01/21)

For clarity, we will call all movies or shows (ie. rows) in this dataset a "program".


### Research Questions

We will use this dataset to answer the following research questions.

1. **Do the 25 longest Netflix programs have an unusually high average imdb rating?**
    - To help us answer this, we will first calculate the average imdb rating of the 25 longest movies $\bar{x}_{longest25}$. Then we will calculate the probability of randomly sampling 25 programs with an average rating that is at least this high (ie. $P(\bar{X}\geq \bar{x}_{longest25}))$.
2. **Do the 30 Netflix programs with the highest Imdb score have an unusually high proportion of documentaries?**
    - To help us answer this, we will first calculate the the proportion of the 30 Netflix programs with the highest imdb score that are documentaries (ie. $\hat{p}_{top30}$). Then we will calculate the probability of randomly sampling 30 programs with a proportion of documentaries that is at least this high (ie. $P(\hat{P}\geq \hat{p}_{top30}))$.


### Imports

````python
import pandas as pd                    # imports pandas and calls the imported version 'pd'
import matplotlib.pyplot as plt        # imports the package and calls it 'plt'
import seaborn as sns # imports the seaborn package with the imported name 'sns'
sns.set() 
import scipy
````

## 1. Data Basics


Read the NetflixOriginals.csv into a dataframe and display the first 5 rows. **You can assume that this dataset has no missing values**.

````python
netflix = pd.read_csv('NetflixOriginals.csv')
netflix.head()
````

## 2.  IMDB Score Population Analysis

In this analysis, we will assume that this full dataset is our population of interest.

### 2.1. Population Mean and Standard Deviation

Calculate the mean and the standard deviation IMDB score of the programs in this population.

````python
netflix['IMDB Score'].mean()
````

````python
netflix['IMDB Score'].std()
````

### 2.2. Population Distribution Shape

Plot the population of IMDB program scores in a histogram and describe the shape of this population distribution.

Make sure you give your plot an appropriate title, x-axis, and y-axis labels.

````python
from scipy.stats import kurtosis, skew
plt.hist(netflix['IMDB Score'])
plt.title('IMDB scores distribution')
plt.xlabel('score')
plt.ylabel('Frequency')
print("Kurtosis: " + str(kurtosis(netflix['IMDB Score'],fisher = False)))
print("Skewness: " + str(skew(netflix['IMDB Score'])))
````

the distribution is left skewed. Leptokurtic

## 3. Sample of the 25 Longest Programs

### 3.1. Create the sample

Create a sample dataframe comprised of the 25 longest programs in this population.

````python
longest = netflix.sort_values(by = "Runtime", ascending = False).head(25)
longest
````

### 3.2. Sample Mean and Standard Deviation

Calculate the mean and the standard deviation IMDB score of the programs in this sample.

````python
longest['IMDB Score'].mean()
````

````python
longest['IMDB Score'].std()
````

### 3.3. Sample Distribution Shape

Plot the population of IMDB program scores in a histogram and describe the shape of this sample distribution.

Make sure you give your plot an appropriate title, x-axis, and y-axis labels.

````python
plt.hist(longest['IMDB Score'])
plt.title('IMDB Score Distribution')
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
print("Kurtosis: " + str(kurtosis(longest['IMDB Score'],fisher = False)))
print("Skewness: " + str(skew(longest['IMDB Score'])))
````

the shape of the distribution is left skewed, Mesokurtic

## 4. Sampling Distribution of Sample Means

Next, let's create a sampling distribution of sample means and analyze it


### 4.1. Create the Sampling Distribution

Specifically, we would like to create a sampling distribution of **sample mean IMDB scores**. The random samples that we are taking the mean IMDB score of should all be of size 25 and be randomly drawn from the population. (Given that this is a sampling distribution, you should know whether we are doing this with or without replacement).

Let's have this sampling distribution be comprised of 5000 sample means.

Create this sampling distribution below.

````python
sample = []
for i in range(5000):
    sample.append(netflix.sample(25)['IMDB Score'].mean())
sample
````

````python
samples = pd.DataFrame(data = sample, columns = ['average'])
samples
````

### 4.2. Sampling Distribution Mean and Standard Deviation

Next, calculate the mean and standard deviation of this sampling distribution of sample means.

````python
samples['average'].mean()
````

````python
samples['average'].std()
````

### 4.3. Sampling Distribution Shape

Next, plot this sampling distribution in a histogram and describe the shape of this sampling distribution. Make sure that you give this histogram an appropriate title, x-axis and y-axis labels.

````python
samples['average'].hist()
plt.title('IMDB Score Distribution')
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
print("Kurtosis: " + str(kurtosis(samples['average'],fisher = False)))
print("Skewness: " + str(skew(samples['average'])))
````

The shape of the distribution is left skewed, Mesokurtic

### 4.4. Approximating Sample Mean Probabilities

Finally, use this sampling distribution to *approximate* the following probability: $P(\bar{X} \geq \bar{x}_{longest25})$

That is, use this sampling distribution to *approximate* the probability of randomly sampling 25 programs with a mean IMDB score that is at least as high as the mean IMDB score of our sample of the 25 longest programs from 3.2.

````python
len(samples[samples['average'] >= longest['IMDB Score'].mean()])/len(samples)
````

## 5.  Netflix Documentary Population Analysis

Again, in this analysis, we are assuming that this full dataset is our population of interest.

### 5.1. Population Proportion

Calculate the proportion of programs in this population that are a documentary.

````python
len(netflix[netflix['Genre'] == 'Documentary'])/len(netflix)
````

## 6. Sample of the 30 Longest Movies

### 6.1. Create the sample

Create a sample dataframe comprised of the 30 programs with the highest IMDB score.

````python
highest = netflix.sort_values(by = 'IMDB Score', ascending = False).head(30)
highest
````

### 6.2. Sample Proportion

Calculate the proportion of programs in this sample that are documentaries.

````python
Doc = len(highest[highest['Genre'] == 'Documentary'])/len(highest)
Doc
````

## 7. Sampling Distribution of Sample Proportions

Next, let's create a sampling distribution of sample proportions and analyze it.


### 7.1. Create the Sampling Distribution

Specifically, we would like to create a sampling distribution of **the proportions of programs in our samples that are documentaries** (ie. $\frac{\mbox{num of documentaries in sample 1}}{\mbox{sample 1 size}}, \frac{\mbox{num of documentaries in sample 2}}{\mbox{sample 2 size}},...,\frac{\mbox{num of documentaries in sample 7000}}{\mbox{sample 7000 size}}$). 

The random samples should all be of size 30 and be randomly drawn from the population. (Given that this is a sampling distribution, you should know whether we should be doing this with or without replacement).

Let's have this sampling distribution be comprised of 7000 sample proportions.

Create this sampling distribution below.

````python
sample = []
for i in range(7000):
    a = netflix.sample(30)
    sample.append(len(a[a['Genre'] == 'Documentary'])/30)
````

````python
samples = pd.DataFrame(data = sample, columns = ['Sample Proportions'])
samples
````

### 7.2. Sampling Distribution Mean and Standard Deviation

Next, calculate the mean and standard deviation of this sampling distribution of sample proportions.

````python
samples['Sample Proportions'].mean()
````

````python
samples['Sample Proportions'].std()
````

### 7.3. Sampling Distribution Shape

Next, plot this sampling distribution in a histogram and describe the shape of this sampling distribution. Make sure that you give this histogram an appropriate title, x-axis and y-axis labels.

````python
samples['Sample Proportions'].hist()
plt.xlabel('Proportions')
plt.ylabel('Frequency')
plt.title('Sampling Distribution of Sample Proportions')
print("Kurtosis: " + str(kurtosis(samples['Sample Proportions'],fisher = False)))
print("Skewness: " + str(skew(samples['Sample Proportions'])))
````

The shape of the sampling distribution is right skewed, Mesokurtic

### 7.4. Approximating Sample Mean Probabilities

Finally, use this sampling distribution to *approximate* the following probability: $P(\hat{P} \geq \hat{p}_{top30})$

That is, use this sampling distribution to *approximate* the probability of randomly sampling 30 programs from this population in which the proportion of documentaries in this sample is at least as high as the proportion we found in 6.2.

````python
len(samples[samples['Sample Proportions'] >= Doc])/len(samples)
````

## <u>Case Study 2</u>: Job Interview Statistics

Suppose that you are on the job market for a data scientist position. After conducting background research, you have discovered that previous data scientists applicants with a similar set of skills and background got job offers for 20% of the data scientists positions that they applied to. So you will assume in this analysis that the probability of you getting hired for any data scientist job that you apply to is **20%**. 

Furthermore let's assume that you get really quick feedback on your applications, and so you know the outcome of one application before applying to another.

Finally, let's assume that you will take the first job offer that you get, and not apply for any more jobs after that.

Let's define a random variable as **X = the number of jobs you apply to until you get a job offer**.

## 8. Job Offer Probabilities

### 8.1. Exactly 10 applications

What is the probability that you have to apply to exactly 10 job applications before getting a job (ie. what is $P(X=10)$)?

````python
from scipy.stats import norm, geom
0.8**9 * 0.2
````

### 8.2. At most 10 applications

What is the probability that you have to apply to at most 10 job applications before getting a job (ie. what is $P(X\leq 10)$)?

````python
prob = 0
for i in range(10):
    prob += 0.8**(i) * 0.2
prob
````

## 9. Job Offer Summary Statistics

### 9.1. Mean Number of Job Applications

On average, how many jobs do you expect to apply for before getting a job?

(Or in other words, suppose we were to conduct this job application "experiment" many, many times where a person with similar background keeps applying to jobs until they get an offer and then records X=the number of applications it took them to get a job offer. What would we expect the average of these many, many X values to be?)

````python
geom.mean(p = 0.2)
````

### 9.2. Median Number of Job Applications

What would you expect the median number of jobs that you would need to apply for before getting a job?

(Or in other words, suppose we were to conduct this job application "experiment" many, many times where a person with similar background keeps applying to jobs until they get an offer and then records X=the number of applications it took them to get a job offer. What would we expect the median of these many, many X values to be?)

````python
geom.median(p = 0.2)
````

## 10. Generating Random Variable Values


### 10.1 Generating 100 Random Variable Values
Generate 100 random values for this **X** random variable. (Or in other words, simulate this "job application experiment" 100 times, and for each experiment give the number of jobs that this person would have to apply to before getting a job.)

````python
X = geom.rvs(p = 0.2, size = 100)
X
````

Put these 100 random variable values into a dataframe and calculate the mean of these values. How close are these values to the mean that you calculated in 9.1?
