In a Simple Linear Regression model,the relationship between two variables: the independent variable x and the dependent variable y. The model is represented by a line:ùëåùëñ=ùõΩ0+ùõΩ1ùë•ùëñ+ùúñùëñ
ùõΩ0= Intercept
ùõΩ1= Slope
œµ = Error term

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the Simple Linear Regression model
beta_0 = 5   # Intercept
beta_1 = 2   # Slope
sigma = 1    # Standard deviation of the error term

# Generate a fixed set of X values (predictor variable)
np.random.seed(0)
X = np.linspace(0, 10, 50)  # 50 values from 0 to 10

# Calculate the expected Y values based on the linear relationship (without noise)
Y_expected = beta_0 + beta_1 * X

# Generate random noise from a normal distribution with mean 0 and standard deviation sigma
error = np.random.normal(0, sigma, size=X.shape)

# Calculate the observed Y values by adding the error term to the expected Y values
Y_observed = Y_expected + error

# Plot the distribution of Y values at a specific X value
X_point = 5
Y_point_expected = beta_0 + beta_1 * X_point
Y_point_dist = np.random.normal(Y_point_expected, sigma, 1000)  # Sample for distribution

# Plotting
plt.figure(figsize=(12, 6))

# Plot the regression line (expected Y values)
plt.plot(X, Y_expected, color="blue", label="Expected Y (Regression Line)")

# Plot the observed Y values (with noise)
plt.scatter(X, Y_observed, color="orange", alpha=0.7, label="Observed Y")

# Plot the distribution of Y at X = 5
plt.hist(Y_point_dist, bins=30, density=True, alpha=0.5, color="green", label="Distribution of Y at X = 5")
x_vals = np.linspace(Y_point_expected - 3*sigma, Y_point_expected + 3*sigma, 100)
plt.plot(x_vals, norm.pdf(x_vals, Y_point_expected, sigma), color="green", linestyle="--")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression: Expected and Observed Y values")
plt.legend()
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate data
# Define the parameters for the true regression model
beta_0 = 5     # Intercept
beta_1 = 2     # Slope
sigma = 1      # Standard deviation of the error term

# Generate a fixed set of X values
np.random.seed(0)
X = np.linspace(0, 10, 50)  # 50 values from 0 to 10

# Calculate the expected Y values (without noise)
Y_expected = beta_0 + beta_1 * X

# Generate random noise from a normal distribution
error = np.random.normal(0, sigma, size=X.shape)

# Calculate the observed Y values by adding noise
Y_observed = Y_expected + error

# Create a DataFrame with the simulated data
data = pd.DataFrame({"X": X, "Y": Y_observed})

# Step 2: Fit a Simple Linear Regression model using statsmodels
# Using Ordinary Least Squares (OLS) regression
model = smf.ols("Y ~ X", data=data).fit()

# Print the model summary to display the fitted coefficients and statistics
print(model.summary())

# Step 3: Visualize the fitted regression line along with the observed data points
plt.figure(figsize=(10, 6))

# Plot observed data points
plt.scatter(data["X"], data["Y"], color="orange", alpha=0.7, label="Observed Y")

# Plot the fitted regression line
X_pred = np.linspace(0, 10, 50)
Y_pred = model.predict(pd.DataFrame({"X": X_pred}))
plt.plot(X_pred, Y_pred, color="blue", label="Fitted Regression Line")

# Labeling
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
````

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>
> _$R^2 = 1 - \frac{\sum_{i=1}^n(Y_i-\hat y)^2}{\sum_{i=1}^n(Y_i-\bar Y)^2}$_
>
> _The visuzation provided in the previous problem can be used to consider $(Y_i-\bar Y)^2$ as the squared distance of the $Y_i$ to their sample average $\bar Y$ as opposed to the squared **residuals** $(Y_i-\hat y)^2$ which is the squared distance of the $Y_i$ to their fitted (predicted) values $Y_i$._

````python
import plotly.express as px

long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

print(smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().summary().tables[1])

# Create a scatter plot with a linear regression trendline
fig = px.scatter(old_faithful[long_wait], x='waiting', y='duration', 
                 title="Old Faithful Geyser Eruptions for short wait times (>"+str(long_wait_limit)+")", 
                 trendline='ols')
fig.show() # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from scipy import stats

# Load and preprocess Old Faithful dataset (assuming it's available as old_faithful)
# For illustration, let's assume old_faithful has columns `waiting` and `duration`
# old_faithful = pd.read_csv('path_to_old_faithful.csv')  # Load your dataset here

# Filter long wait times based on the criterion provided
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit
long_wait_data = old_faithful[long_wait]

# 1. Bootstrap Sampling and Fitting
bootstrap_slope_coefficients = []
n_bootstraps = 1000  # Number of bootstrap samples

for _ in range(n_bootstraps):
    bootstrap_sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    fitted_model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slope_coefficients.append(fitted_model.params[1])  # Collect the slope

# Visualize Bootstrap Sampling Distribution
plt.hist(bootstrap_slope_coefficients, bins=30, color='lightblue', edgecolor='black')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.title('Bootstrap Sampling Distribution of Slope Coefficients')
plt.show()

# Calculate the 95% confidence interval for the bootstrapped slope coefficients
ci_lower, ci_upper = np.percentile(bootstrap_slope_coefficients, [2.5, 97.5])
print(f"95% Bootstrapped Confidence Interval for the Slope: [{ci_lower}, {ci_upper}]")

# 2. Simulated Sampling under Null Hypothesis (No Linear Association)
simulated_slope_coefficients = []
n_simulations = 1000  # Number of simulations

for _ in range(n_simulations):
    simulated_data = long_wait_data.copy()
    simulated_data['duration'] = 1.65 + 0 * simulated_data['waiting'] + \
                                 stats.norm.rvs(loc=0, scale=0.37, size=len(long_wait_data))
    simulated_model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slope_coefficients.append(simulated_model.params[1])  # Collect the slope

# Visualize Simulated Sampling Distribution under Null Hypothesis
plt.hist(simulated_slope_coefficients, bins=30, color='lightcoral', edgecolor='black')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.title('Simulated Sampling Distribution of Slope Coefficients (Null Hypothesis)')
plt.show()

# 3. Check if 0 is within Bootstrapped Confidence Interval
contains_zero = ci_lower <= 0 <= ci_upper
print(f"Does the 95% bootstrapped confidence interval contain 0? {contains_zero}")

# Calculate simulated p-value for the null hypothesis
observed_slope = smf.ols('duration ~ waiting', data=long_wait_data).fit().params[1]
simulated_p_value = (np.abs(simulated_slope_coefficients) >= np.abs(observed_slope)).mean()
print(f"Simulated p-value: {simulated_p_value}")
````

### Current Course Project Capability Level

At this point in the course you should be able to do a **Simple Linear Regression** analysis for data from the Canadian Social Connection Survey data
    
1. Create and test a **null hypothesis** of no linear association "on average" for a couple of columns of interest in the Canadian Social Connection Survey data using **Simple Linear Regression**

2. Use the **residuals** of a fitted **Simple Linear Regression** model to diagnostically assess some of the assumptions of the analysis

3. Use an **indicator variable** based **Simple Linear Regression** model to compare two groups from the Canadian Social Connection Survey data

4. Compare and contrast the results of an **indicator variable** based **Simple Linear Regression** model to analyses based on a **permutation test** and a **bootstrapped confidence interval**
