```python
"""
1. What is the role of the 'else' block in a try-except statement? Provide an example
scenario where it would be useful.

The else condition in a try-except statement is useful when the try condition doesnâ€™t raise an exception. However, it must follow all the except conditions.


prices = { 'Pen' : 10, 'Pencil' : 5, 'Notebook' : 25}
item = input('Get price of:')
try:
    print(f'The price of {item} is {prices[item]}')
except KeyError:
    print(f'The price of {item} is not known')
else:
    print(f'There is no error in the statement')
    
"""
```
