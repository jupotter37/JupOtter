```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f'Using device: {device}')

# Load data
data = pd.read_pickle('./merge/BTCUSDT.pkl')  # Assuming it contains the time series data
# Use all features: open, high, low, close, volume, trades
features = data[['open', 'high', 'low', 'close', 'volume', 'trades']].dropna()[:10000]

# Calculate returns based on the 'close' price and fill NaN values with 0
returns = data['close'].pct_change().fillna(0)[:10000]  # Calculate percentage change and fill NaN with 0
features = features.iloc[:10000]  # Align features with returns length

# Add returns to the features as an additional feature
features['returns'] = returns.values

# Data normalization
feature_scaler = MinMaxScaler(feature_range=(0, 1))
target_scaler = MinMaxScaler(feature_range=(0, 1))

features_scaled = feature_scaler.fit_transform(features)
returns_scaled = target_scaler.fit_transform(returns.values.reshape(-1, 1))

# Split dataset - 80% for training, 20% for testing
train_size = int(len(features_scaled) * 0.8)
train_features = features_scaled[:train_size]
test_features = features_scaled[train_size:]

train_returns = returns_scaled[:train_size]
test_returns = returns_scaled[train_size:]

# Prepare training data
look_back = 60  # Use the past 60 days to predict the next day's return

def create_sequences(features, target, look_back):
    X, y = [], []
    for i in range(look_back, len(features)):
        X.append(features[i-look_back:i, :])  # All features for the past 'look_back' days
        y.append(target[i, 0])  # Predicting the return of the next day
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_features, train_returns, look_back)
X_test, y_test = create_sequences(test_features, test_returns, look_back)

# Convert to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float32).to(device)
y_train = torch.tensor(y_train, dtype=torch.float32).to(device)
X_test = torch.tensor(X_test, dtype=torch.float32).to(device)
y_test = torch.tensor(y_test, dtype=torch.float32).to(device)

# Create data loader
batch_size = 32
train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Define LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size=7, hidden_size=50, num_layers=2, output_size=1):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])  # Use output of the last time step
        return out

# Instantiate model, define loss function and optimizer
model = LSTMModel(input_size=7).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
epochs = 100
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()  # Zero the gradients
        outputs = model(X_batch)  # Forward pass
        loss = criterion(outputs, y_batch.unsqueeze(1))  # Compute loss
        loss.backward()  # Backward pass
        optimizer.step()  # Update parameters
        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}")

# Make predictions
model.eval()
with torch.no_grad():
    predicted_train_returns = model(X_train).cpu().numpy()
    predicted_test_returns = model(X_test).cpu().numpy()

# Inverse transform predicted and actual values
y_train_actual = target_scaler.inverse_transform(y_train.cpu().numpy().reshape(-1, 1))
predicted_train_returns = target_scaler.inverse_transform(predicted_train_returns)
y_test_actual = target_scaler.inverse_transform(y_test.cpu().numpy().reshape(-1, 1))
predicted_test_returns = target_scaler.inverse_transform(predicted_test_returns)

# Visualize the results
plt.figure(figsize=(14, 6))

# Training set predictions
plt.plot(range(look_back, train_size), y_train_actual, color='blue', label='Actual Returns (Training Set)')
plt.plot(range(look_back, train_size), predicted_train_returns, color='red', label='Predicted Returns (Training Set)')

# Testing set predictions
plt.plot(range(train_size + look_back, train_size + look_back + len(y_test_actual)), y_test_actual, color='green', label='Actual Returns (Testing Set)')
plt.plot(range(train_size + look_back, train_size + look_back + len(predicted_test_returns)), predicted_test_returns, color='orange', label='Predicted Returns (Testing Set)')

plt.xlabel('Time')
plt.ylabel('Returns')
plt.legend()
plt.show()
```

```python
import numpy as np 
import pandas as pd
np.random.seed(42)

def take_profitANDstop_loss(close, order_volume, up:float, low:float):
    mask = np.triu(np.ones((len(close),len(close)), dtype=bool))
    condition_up = np.where(mask, np.tile(close*up,[len(close), 1]).T, np.nan) #止盈矩阵，每一列都是行时刻下单对应止盈止损点，且是上三角阵，即只判断下单后的，下单前都为0
    condition_low = np.where(mask, np.tile(close*low,[len(close), 1]).T, np.nan) #止损矩阵，每一列都是行时刻下单对应止盈止损点，且是上三角阵，即只判断下单后的，下单前都为0
    exit_time_up = pd.DataFrame(condition_up <= np.array(close)).idxmax(axis=1).replace(0, np.inf)  #每个时刻close与所有时刻止盈点比较，找到第一次大于止损点时刻，得到Series
    exit_time_low = pd.DataFrame(condition_low >= np.array(close)).idxmax(axis=1).replace(0, np.inf) #每个时刻close与所有时刻止损点比较，找到第一次小于止损点时刻，得到Series
    exit_time = exit_time_up.combine(exit_time_low, min)#先止盈还是先止损
    pos_matrix = pd.DataFrame(np.tile(range(len(close)), [len(close),1])).T #位置矩阵，第n行为全n 行向量
    tile_pos = pd.DataFrame(np.tile(order_volume,[len(close),1])) #初始订单量拓展矩阵，每一列是列时刻订单量的常向量
    adjust = np.tril(tile_pos * (pos_matrix == exit_time)).sum(1) #通过广播将位置矩阵与平仓时刻比较，每一列是 列时刻订单在行时刻平仓数量的列向量, 将行时刻需要平仓数量加总，即可得到每时刻需平仓量
    order_volume = pd.DataFrame(order_volume - adjust)
    order_volume.index = close.index
    return order_volume


close = pd.read_parquet('BTCUSDT.parquet')['close'][:10000]
up = 1.00001
low = 0.99999
order_volume = np.random.rand(len(close))
take_profitANDstop_loss(close, order_volume, up, low)
```

```python
import numpy as np
import pandas as pd
import time

np.random.seed(42)

def take_profitANDstop_loss_optimized(close, order_volume, up: float, low: float):
    close_array = close.to_numpy()
    len_close = len(close_array)
    
    # 计算止盈和止损点
    condition_up = close_array * up
    condition_low = close_array * low

    # 初始化平仓数量调整数组
    adjust = np.zeros(len_close)

    # 使用更高效的方法进行查找
    exit_times = np.full(len_close, np.inf)
    
    for i in range(len_close):
        up_condition_met = np.searchsorted(close_array[i:], condition_up[i], side='left') + i
        low_condition_met = np.searchsorted(close_array[i:], condition_low[i], side='right') + i

        if up_condition_met < len_close and close_array[up_condition_met] >= condition_up[i]:
            exit_time_up = up_condition_met
        else:
            exit_time_up = np.inf

        if low_condition_met < len_close and close_array[low_condition_met] <= condition_low[i]:
            exit_time_low = low_condition_met
        else:
            exit_time_low = np.inf

        # 选择止盈或止损中较早发生的时间
        exit_times[i] = min(exit_time_up, exit_time_low)

    # 累加平仓数量
    valid_exit_times = exit_times[exit_times < len_close].astype(int)
    np.add.at(adjust, valid_exit_times, order_volume[exit_times < len_close])

    # 计算调整后的订单量
    order_volume_adjusted = order_volume - adjust

    # 构建结果 DataFrame
    order_volume_df = pd.DataFrame(order_volume_adjusted, index=close.index)
    
    return order_volume_df

start_time = time.time()
close = pd.read_parquet('BTCUSDT.parquet')['close'][:100000]
up = 1.00001
low = 0.99999
order_volume = np.random.rand(len(close))
take_profitANDstop_loss_optimized(close, order_volume, up, low)
end_time = time.time()
print("Execution time:", end_time - start_time)
```

```python
import numpy as np
import pandas as pd
import time

np.random.seed(42)

def take_profitANDstop_loss_optimized(close, order_volume, up: float, low: float):
    # 使用输入的 numpy array 而不是转换
    close_array = close if isinstance(close, np.ndarray) else close.to_numpy()
    len_close, num_assets = close_array.shape
    
    # 计算止盈和止损点
    condition_up = close_array * up
    condition_low = close_array * low

    # 初始化平仓数量调整数组
    adjust = np.zeros((len_close, num_assets))

    # 使用广播和 NumPy 向量化操作来找到止盈和止损的时间
    up_mask = close_array >= condition_up
    low_mask = close_array <= condition_low

    exit_times_up = np.argmax(up_mask, axis=0)
    exit_times_low = np.argmax(low_mask, axis=0)

    # 如果没有找到满足条件的位置，将其设置为 len_close
    exit_times_up[~up_mask.any(axis=0)] = len_close
    exit_times_low[~low_mask.any(axis=0)] = len_close

    # 找到止盈和止损中较早的时间
    exit_times = np.minimum(exit_times_up, exit_times_low)

    # 累加平仓数量，使用 NumPy 的广播来加速
    valid_exit_mask = exit_times < len_close
    for j in range(num_assets):
        if valid_exit_mask[j]:
            np.add.at(adjust[:, j], exit_times[j], order_volume[exit_times[j], j])

    # 计算调整后的订单量
    order_volume_adjusted = order_volume - adjust

    # 构建结果 DataFrame
    if isinstance(close, pd.DataFrame):
        order_volume_df = pd.DataFrame(order_volume_adjusted, index=close.index, columns=close.columns)
    else:
        order_volume_df = order_volume_adjusted
    
    return order_volume_df

start_time = time.time()
close = np.random.rand(400000,500)
up = 1.00001
low = 0.99999
order_volume = np.random.rand(len(close),500)
take_profitANDstop_loss_optimized(close, order_volume, up, low)
end_time = time.time()
print("Execution time:", end_time - start_time)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f'Using device: {device}')

# Load data
data = pd.read_pickle('./merge/BTCUSDT.pkl')  # Assuming it contains the time series data
data['return'] = data['close'].pct_change().fillna(0)  # Calculate returns and fill NaN values with 0

# Use all features: open, high, low, close, volume, trades, return
features = data[['open', 'high', 'low', 'close', 'volume', 'trades', 'taker_base_volume', 'return']][:100000]

# Clean outliers using 3 times standard deviation before splitting the data
for column in features.columns:
    mean = features[column].mean()
    std = features[column].std()
    upper_limit = mean + 3 * std
    lower_limit = mean - 3 * std
    features[column] = features[column].clip(lower=lower_limit, upper=upper_limit)

# Data normalization
feature_scaler = MinMaxScaler(feature_range=(0, 1))
target_scaler = MinMaxScaler(feature_range=(0, 1))

features_scaled = feature_scaler.fit_transform(features)
returns_scaled = target_scaler.fit_transform(data['return'].values.reshape(-1, 1))

# Split dataset - 80% for training, 20% for testing
train_size = int(len(features_scaled) * 0.8)
train_features = features_scaled[:train_size]
test_features = features_scaled[train_size:]

train_returns = returns_scaled[:train_size]
test_returns = returns_scaled[train_size:]

# Prepare training data
look_back = 60  # Use the past 60 days to predict the next day's return

def create_sequences(features, target, look_back):
    X, y = [], []
    for i in range(look_back, len(features)):
        X.append(features[i-look_back:i, :])  # All features for the past 'look_back' days
        y.append(target[i, 0])  # Predicting the return of the next day
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_features, train_returns, look_back)
X_test, y_test = create_sequences(test_features, test_returns, look_back)

# Convert to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float32).to(device)
y_train = torch.tensor(y_train, dtype=torch.float32).to(device)
X_test = torch.tensor(X_test, dtype=torch.float32).to(device)
y_test = torch.tensor(y_test, dtype=torch.float32).to(device)

# Create data loader
batch_size = 64
train_dataset = TensorDataset(X_train, y_train)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Define a more complex LSTM model
class ComplexLSTMModel(nn.Module):
    def __init__(self, input_size=7, hidden_size=100, num_layers=3, output_size=1, dropout_rate=0.3):
        super(ComplexLSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout_rate)
        self.fc1 = nn.Linear(hidden_size, 50)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(50, output_size)
    
    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc1(out[:, -1, :])  # Take output of the last time step
        out = self.relu(out)
        out = self.fc2(out)
        return out

# Instantiate model, define loss function and optimizer
model = ComplexLSTMModel(input_size=7, hidden_size=100, num_layers=3, output_size=1, dropout_rate=0.3).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
epochs = 20  # Increase the number of epochs
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()  # Zero the gradients
        outputs = model(X_batch)  # Forward pass
        loss = criterion(outputs, y_batch.unsqueeze(1))  # Compute loss
        loss.backward()  # Backward pass
        optimizer.step()  # Update parameters
        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}")

# Make predictions
model.eval()
with torch.no_grad():
    predicted_train_returns = model(X_train).cpu().numpy()
    predicted_test_returns = model(X_test).cpu().numpy()

# Inverse transform predicted and actual values
y_train_actual = target_scaler.inverse_transform(y_train.cpu().numpy().reshape(-1, 1))
predicted_train_returns = target_scaler.inverse_transform(predicted_train_returns)
y_test_actual = target_scaler.inverse_transform(y_test.cpu().numpy().reshape(-1, 1))
predicted_test_returns = target_scaler.inverse_transform(predicted_test_returns)

# Visualize the results
plt.figure(figsize=(14, 6))

# Training set predictions
plt.plot(range(look_back, train_size), y_train_actual, color='blue', label='Actual Returns (Training Set)')
plt.plot(range(look_back, train_size), predicted_train_returns, color='red', label='Predicted Returns (Training Set)')

# Testing set predictions
plt.plot(range(train_size + look_back, train_size + look_back + len(y_test_actual)), y_test_actual, color='green', label='Actual Returns (Testing Set)')
plt.plot(range(train_size + look_back, train_size + look_back + len(predicted_test_returns)), predicted_test_returns, color='orange', label='Predicted Returns (Testing Set)')

plt.xlabel('Time')
plt.ylabel('Returns')
plt.legend()
plt.show()
```
