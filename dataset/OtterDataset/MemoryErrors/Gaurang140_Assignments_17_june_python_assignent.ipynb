
**Q1. What is the role of try and exception block?**
- **Ans:** The `try` and `except` blocks are used for exception handling in Python. They allow you to write code that may raise exceptions and handle those exceptions gracefully, preventing the program from crashing.


---
---



**Q2. What is the syntax for a basic try-except block?**
- **Ans:** 
```python
try:
    # code that may raise an exception
except ExceptionType:
    # code to handle the exception
```

---
---

**Q3. What happens if an exception occurs inside a try block and there is no matching except block?**
- **Ans:** If an exception occurs inside a try block and there's no matching `except` block, the program will terminate and display an error message indicating the unhandled exception.



---
---

**Q4. What is the difference between using a bare except block and specifying a specific exception type?**
- **Ans:** A bare `except` block catches all types of exceptions, making it harder to diagnose issues. Specifying a specific exception type allows you to handle specific errors and ignore others, making the code more robust.

**Q5. Can you have nested try-except blocks in Python? If yes, then give an example.**
- **Ans:** Yes, you can have nested `try-except` blocks. Example:
```python
try:
    try:
        # inner try block
    except ExceptionTypeInner:
        # handle inner exception
except ExceptionTypeOuter:
    # handle outer exception
```

---
---

**Q6. Can we use multiple exception blocks, if yes then give an example.**
- **Ans:** Yes, you can use multiple `except` blocks to handle different types of exceptions. Example:
```python
try:
    # code that may raise an exception
except ExceptionType1:
    # handle ExceptionType1
except ExceptionType2:
    # handle ExceptionType2
```

**Q8. Write code for the following given scenario and add try-exception block to it.**

- a. Program to divide two numbers:
```python
try:
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))
    result = numerator / denominator
except ZeroDivisionError:
    print("Cannot divide by zero.")
except ValueError:
    print("Invalid input. Please enter valid numbers.")
except Exception as e:
    print("An error occurred:", e)
else:
    print("Result:", result)
```
---
- b. Program to convert a string to an integer:
```python
try:
    num_str = input("Enter a number: ")
    num = int(num_str)
except ValueError:
    print("Invalid input. Not a valid number.")
else:
    print("Number:", num)
```
---
- c. Program to access an element in a list:
```python
try:
    my_list = [1, 2, 3]
    index = int(input("Enter index: "))
    value = my_list[index]
except IndexError:
    print("Index out of range.")
else:
    print("Value at index:", value)
```
---
- d. Program to handle a specific exception:
```python
try:
    num = int(input("Enter a positive number: "))
    if num <= 0:
        raise ValueError("Number must be positive.")
except ValueError as ve:
    print("Error:", ve)
```
---
- e. Program to handle any exception:
```python
try:
    # code that may raise an exception
except Exception as e:
    print("An error occurred:", e)
```
