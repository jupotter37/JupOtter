```python
# Ans 3
# Errors defined in ArithmeticError class are:
# ZeroDivision Error, OverfloError, FloatingPointerError, ValueError

# ZeroDivision Error
try:
    7/0
except ZeroDivisionError as r:
    print("Dividing by zero :(    :", r)

# OverFlowError
j = 5.0
try:
    for i in range (1, 1000):
        j = j**i
    print (j)
except OverflowError as e:
    print ("Overflow error happened")
    print (f" {e}, {e.__class__}")
```

```python
# Ans 4
# The LookupError exception is the base class for errors raised when something can’t be found. It is used when a key or index used on a mapping or sequence is invalid 1.

# KeyError and IndexError are two examples of exceptions that inherit from the LookupError class.

# A KeyError is raised when you try to access a key that does not exist in a dictionary . For example, if you have a dictionary my_dict = {'a': 1, 'b': 2} and you try to access the value of the key ‘c’, you will get a KeyError because ‘c’ is not in the dictionary.

# An IndexError is raised when you try to access an index that is out of range in a sequence like a list or tuple . For example, if you have a list my_list = [1, 2, 3] and you try to access the value at index 3 (which does not exist), you will get an IndexError.
```

```python
# Ans 5

# The ImportError exception is raised when a module or member of a module cannot be imported
try:
    import my_module
except ImportError:
    print("The module could not be imported.")
    
    
# The ModuleNotFoundError exception is raised when a module cannot be found.
# This can happen if the module does not exist, is not installed, or is not in the correct location.
try:
    import my_module
except ModuleNotFoundError:
    print("The module could not be found.")
```
