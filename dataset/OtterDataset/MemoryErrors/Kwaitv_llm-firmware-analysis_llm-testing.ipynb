```python
# Use a pipeline as a high-level helper
from transformers import pipeline

#obj_file = "stream-disasm.txt" 
obj_file = "stream-decomp.txt" 
dis = ""
with open(obj_file, 'r') as file:
    dis = file.read()


pipe = pipeline("text-generation", 
                model="unsloth/llama-3-8b-Instruct-bnb-4bit")
```

```python
def query(text, prev):
    message = []
    for i in text:
        message.append({"role": "user", "content": i})
    output = pipe(message)
    return output


system = [
    {"role" : "system", 
     "content" : "you are part of a system that takes the output of a decompiler in the form of c code and tries to assist in analyze the original code it came from"
    }
]
queries = [
    "{} what does this c code do?".format(dis),
    "what inputs parameters does the function take as input?",
    "what type of data does the function return?",
]
```

```python
output = query(queries[0], system)
print(output[-1]['generated_text'][-1]['content'])
```

```python
output2 = query(queries[1], output)
print(output2[-1]['generated_text'][-1]['content'])
```
