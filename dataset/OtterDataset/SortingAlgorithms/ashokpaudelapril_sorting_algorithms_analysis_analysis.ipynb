```python
import matplotlib.pyplot as plt
```

```python
# Updated data from your sorting results
sizes = [10, 50, 100, 250, 500, 1000]
bubble_steps = [0.073, 1.860, 7.301, 47.285, 186.214, 748.505]
selection_steps = [0.052, 1.269, 5.047, 31.370, 125.239, 500.496]
insertion_steps = [0.067, 1.361, 4.873, 32.809, 123.517, 499.119]

```

```python
# Plotting the data
plt.figure(figsize=(12, 6))
```

```python

plt.plot(sizes, bubble_steps, marker='o', label='Bubble Sort', color='blue')
plt.plot(sizes, selection_steps, marker='o', label='Selection Sort', color='orange')
plt.plot(sizes, insertion_steps, marker='o', label='Insertion Sort', color='green')

# Adding titles and labels
plt.title('Sorting Algorithms Performance')
plt.xlabel('Array Size')
plt.ylabel('Number of Steps')

# Set x-axis ticks
plt.xticks(sizes)  # Ensure the x-axis only shows the array sizes

# Adjust y-axis limits to zoom in (modify these limits as necessary)
plt.ylim(1, 800)  # Adjust this limit to zoom in on the desired range

plt.legend()
plt.grid(True)

plt.savefig('sorting_algorithms_performance.png', format='jpg', dpi=300)  # Save as PNG file with high resolution

# Show plot
plt.tight_layout()
plt.show()
```
