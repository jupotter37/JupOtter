```python
# Import Data from CSV file
import numpy as np
import pandas as pd
dataset = file_path = '/content/Online retails.xlsx'
dataset = pd.read_excel(file_path)
dataset = dataset.replace(np.nan, '', regex=True)
dataset
```

```python
dataset
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,u]) for u in range(0, 19)])
```

```python
retail_series  = pd.DataFrame(pd.Series(transactions))
retail_series
retail_series.columns = ["business"]
retail_series
```

```python
X = retail_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X, min_support=0.005,max_len=3,use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1,num_itemsets=True)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)
```
