```python
import cohere
import pandas as pd
```

```python
cohere_api_key = '050aYyOJYnTlqOAuCMI******'
co = cohere.Client(cohere_api_key)
```

```python
def reframe_question(question):
    response = co.generate(
        model='command-r-plus',
        prompt=f"Rephrase the following question: {question}",
        max_tokens=50,
        temperature=0.7
    )
    reframed_question = response.generations[0].text.strip()
    return reframed_question
```

```python
dataset_path = '/content/questions_task.xlsx'
questions_df = pd.read_excel(dataset_path)
questions_df
```

```python
print("Welcome to the chatbot! Type 'exit' if you want to leave..")
```

```python
while True:
    print("\nAvailable questions:")
    for i, question in enumerate(questions_df['questions'], 1):
        print(f"{i}. {question}")

    user_input = input("\nEnter the question number you want to reframe or type 'exit' to leave: ").strip()

    if user_input.lower() == 'exit':
        print(" Thank you for using our bot Good day!!")
        break

    # user will input is a number
    if not user_input.isdigit() or int(user_input) < 1 or int(user_input) > len(questions_df):
        print("Incorrect input. Please enter a valid question number or 'exit' to leave.")
        continue

    # get the selected que.. based on user input
    question_number = int(user_input)
    selected_question = questions_df['questions'][question_number - 1]

    reframed = reframe_question(selected_question)
    print(f"\nOriginal question in dataset: {selected_question}")
    print(f"--------------------------New Reframed question:----> {reframed}")
```
