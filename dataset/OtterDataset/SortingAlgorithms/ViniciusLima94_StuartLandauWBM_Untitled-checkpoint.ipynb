```python
"""\
Builds and/or visualizes the channel positions file, used by the spike sorting algorithm.

Usage: Fill the three variables and execute.
"""

import json
import os

import matplotlib.pyplot as plt
import numpy as np

rat = "jp"            # Config name to use. Should be a rat name.
extra_channels = 17   # How many non-electrode data channels you use (Ephys "additional channels" + Ephys BNC analog inputs)
save = True           # Set to true to save the positions file. THIS WILL OVERWRITE ANY FILE WITH THE SAME NAME. Set to false to only visualize the channel positions.


def visualize_positions(pos_array):
    n_plots = len(pos_array) // pos_array.shape[0]
    if n_plots == 1:
        plt.scatter(pos_array[:,0], pos_array[:,1])
        for i in range(len(pos_array)):
            plt.annotate(i, pos_array[i,:2])
    else:
        axs = []
        for p in range(1, n_plots):
            axs.append(plt.subplot(3,4,p))
        for i, point in enumerate(pos_array):
            if point[2] < 0.5:
                continue
            ax = axs[int(point[2])-1]
            ax.scatter([point[0]], [point[1]], c='b')
            ax.annotate(i, point[:2])
    plt.show()

    
def _get_sites(deltas, sites="even"):
    a0, a1 = -29.5 / 15, 20
    if sites == "odd":
        a0 = - a0
    return np.stack(( a0 * deltas, a1 * deltas, np.zeros(len(deltas))), axis=1)

n_chaines = 32
n_even = n_chaines // 2
n_odd = n_even - 1
n_chunks = 4
    
# Defining physical electrode sites on one shank.
even_sites = _get_sites(np.arange(15, 0, -15 / n_even), sites="even")
odd_sites =  _get_sites(np.arange(14, 0, -14 / n_odd), sites="odd")
sites = [[even_sites[i], odd_sites[i]] for i in range(len(odd_sites))] + [[even_sites[-1], (0, -5, 0)]]
sites_on_shank_positions = np.reshape(np.array(sites), (n_chaines, 3))

visualize_positions(sites_on_shank_positions)


positions = [] 

for i in range(n_chunks * 3):
    if i == 0:
        positions +=[  np.copy(sites_on_shank_positions) + np.array([[0, 0, 1]]) ]
    else:
        positions +=[np.copy(positions[-1]) + np.array([[0, 0, 1]])] 


positions = np.concatenate(positions)

visualize_positions(positions)


# Adjusting positions, accounting for weird numbering of channels on the shanks. Here adjusted_positions[a] = (x,y) means that channel a is at coordinates (x,y)
with open(os.path.abspath(os.path.join(_file, f"../../config{rat}.json"))) as f:
    config = json.load(f)
channels_by_region = config["channels_by_region"]
channel_sites = channels_by_region["NR"] + channels_by_region["PFC"] + channels_by_region["HPC"] + list(range(192,192+extra_channels))     # Use the same region order as above

adjusted_positions = np.zeros((len(channel_sites), 3))
for i in range(len(channel_sites)):
    adjusted_positions[channel_sites[i]] = positions[i]

visualize_positions(adjusted_positions)

# Saving the array
if save:
    np.save(os.path.abspath(os.path.join(_file, os.path.pardir, f"positions{rat}_3D.npy")), adjusted_positions)
    print("Positions saved.")
```
