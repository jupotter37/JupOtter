# Data Structure - Sorting Techniques

Sorting refers to arranging data in a particular format. Sorting algorithm specifies the way to arrange data in a particular order. Most common orders are in numerical or lexicographical order.

The importance of sorting lies in the fact that data searching can be optimized to a very high level, if data is stored in a sorted manner. Sorting is also used to represent data in more readable formats.

Adaptive and Non-Adaptive Sorting Algorithm
A sorting algorithm is said to be adaptive, if it takes advantage of already 'sorted' elements in the list that is to be sorted. That is, while sorting if the source list has some element already sorted, adaptive algorithms will take this into account and will try not to re-order them.

A non-adaptive algorithm is one which does not take into account the elements which are already sorted. They try to force every single element to be re-ordered to confirm their sortedness.

 # Important Terms
 
Some terms are generally coined while discussing sorting techniques, here is a brief introduction to them −

**Increasing Order**
A sequence of values is said to be in increasing order, if the successive element is greater than the previous one. For example, 1, 3, 4, 6, 8, 9 are in increasing order, as every next element is greater than the previous element.

**Decreasing Order**
A sequence of values is said to be in decreasing order, if the successive element is less than the current one. For example, 9, 8, 6, 4, 3, 1 are in decreasing order, as every next element is less than the previous element.

**Non-Increasing Order**
A sequence of values is said to be in non-increasing order, if the successive element is less than or equal to its previous element in the sequence. This order occurs when the sequence contains duplicate values. For example, 9, 8, 6, 3, 3, 1 are in non-increasing order, as every next element is less than or equal to (in case of 3) but not greater than any previous element.

**Non-Decreasing Order**
A sequence of values is said to be in non-decreasing order, if the successive element is greater than or equal to its previous element in the sequence. This order occurs when the sequence contains duplicate values. For example, 1, 3, 3, 6, 8, 9 are in non-decreasing order, as every next element is greater than or equal to (in case of 3) but not less than the previous one.

Bubble Sort is a simple sorting algorithm that repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. The pass through the list is repeated until the list is sorted. It is called "Bubble Sort" because the smaller elements "bubble" to the top of the list as the algorithm progresses.

# Pseudocode

```python
procedure bubbleSort(list: array of items)
   loop = list.count;
   
   for i = 0 to loop-1 do:
      swapped = false
		
      for j = 0 to loop-1 do:
         /* compare the adjacent elements */   
         if list[j] > list[j+1] then
            /* swap them */
            swap(list[j], list[j+1])		 
            swapped = true
         end if
      end for
      
      /* If no number was swapped, that means 
      the array is sorted now, break the loop. */
      
      if (not swapped) then
         break
      end if
   end for
end procedure

return list
```


list is an array of items that you want to sort.

loop is initialized to the number of elements in the list.

We have two nested loops. The outer loop runs from 0 to loop - 1. It keeps track of the number of passes through the list.

Inside the outer loop, we initialize swapped to false. This variable will help us determine whether any swaps were made during a pass.

The inner loop runs from 0 to loop - 1. It compares adjacent elements in the list and swaps them if they are out of order.

If a swap is made, we set swapped to true to indicate that there was a swap during this pass.

After the inner loop completes, we check the value of swapped. If it is still false, it means no swaps were made during this pass, and the list is already sorted. In this case, we break out of the outer loop early to optimize the algorithm.

Finally, we return the sorted list.

# Analysis:

The time complexity of the Bubble Sort algorithm is O(n^2) in the worst and average cases, where 'n' is the number of elements in the list.

This is because, in the worst case, the algorithm has to make n passes through the list, and in each pass, it may perform n comparisons and swaps.

The best-case time complexity is O(n) when the list is already sorted, and no swaps are needed. However, the worst-case and average-case time complexities make Bubble Sort inefficient for large lists. 

It is primarily used for educational purposes and not for practical sorting of large datasets.

Bubble Sort Analysis in Simple Terms:

Imagine you have a list of numbers, and you want to sort them in ascending order using Bubble Sort.

You start by comparing the first two numbers in the list.

If the first number is larger than the second number, you swap them. If not, you leave them as they are.

Then, you move to the next pair of numbers and do the same comparison and swap if needed.

You keep doing this for all pairs of adjacent numbers in the list, moving through the entire list.

After the first pass, the largest number will "bubble up" to the end of the list.

You repeat this process for the remaining unsorted part of the list (excluding the last element, which is already in its correct place).

You keep doing these passes until no more swaps are needed in a pass.

# Now, let's break down the analysis:

Best Case: The best case is when the list is already perfectly sorted. In this case, Bubble Sort will go through the list once to check if any swaps are needed (but no swaps will be needed because it's already sorted). So, it's like checking once, which takes a short time. This is the fastest scenario, and we say Bubble Sort has a time complexity of O(n) in the best case.

Worst Case: The worst case is when the list is in reverse order, meaning the largest number is at the beginning and the smallest is at the end. In this case, Bubble Sort will keep making passes through the list, swapping adjacent elements, until it reaches the end. This requires many passes and many comparisons. We say Bubble Sort has a time complexity of O(n^2) in the worst case because, for a list of n elements, it might take n passes with n comparisons in each pass.

Average Case: On average, if the list is randomly ordered, Bubble Sort will take about as long as the worst-case scenario, which is O(n^2).

# Selection Sort

Selection Sort is another simple comparison-based sorting algorithm. It divides the input list into two parts: the sorted part on the left and the unsorted part on the right. In each pass, Selection Sort finds the minimum (or maximum, depending on the sorting order) element from the unsorted part and moves it to the beginning of the sorted part. It repeats this process until the entire list is sorted.

````python
def selectionSort(array):
    for i in range(len(array) - 1):
        min_idx = i
        
        # Find the index of the minimum element in the unsorted part
        for idx in range(i + 1, len(array)):
            if array[idx] < array[min_idx]:
                min_idx = idx
        
        # Swap the minimum element with the current element at index i
        array[i], array[min_idx] = array[min_idx], array[i]
    
    return array
````

**Explanation:**

The selectionSort function takes an array as input and sorts it in ascending order using Selection Sort.

The outer loop runs from 0 to len(array) - 2. It represents the current position where the minimum element should be placed. Since the last element is already in its correct place when the rest of the list is sorted, we don't need to consider it.

min_idx is initialized to i, which represents the current position where the minimum element is expected to be.

The inner loop starts from i + 1 and iterates through the unsorted part of the list. It compares each element with the element at min_idx, and if a smaller element is found, it updates min_idx to the index of that element.

After finding the minimum element in the unsorted part, it swaps the minimum element with the element at the current position i, effectively moving the minimum element to its correct position.

This process repeats until the entire list is sorted.

Finally, the sorted array is returned.

#  Analysis:

The time complexity of Selection Sort is always O(n^2), where 'n' is the number of elements in the array. This is because, in each pass, it finds the minimum element through a linear scan of the unsorted part and then performs a swap operation. Since it performs these operations n times (n-1 passes in the worst case), the overall time complexity is quadratic.

Selection Sort is not very efficient for large lists, but it has the advantage of making fewer swaps compared to Bubble Sort. It's a simple sorting algorithm that is useful for educational purposes and for small datasets where efficiency is not critical.

# Insertion Sort

Like Selection Sort, in this algorithm, we segment the list into sorted and unsorted parts. Then we iterate over the unsorted segment and insert the element from this segment into the correct position in the sorted list.

Insertion Sort is a simple sorting algorithm that builds the final sorted array one item at a time. It works by repeatedly taking the next unsorted item and "inserting" it into its correct position within the already sorted part of the array. It's similar to how we sort playing cards in our hand.

````python
def insertionSort(array):
    for i in range(1, len(array)):
        key = array[i]
        j = i - 1
        
        # Move elements of array[0..i-1] that are greater than key
        # to one position ahead of their current position
        while array[j] > key and j >= 0:
            array[j + 1] = array[j]
            j -= 1
        
        # Place key at its correct position in the sorted part of the array
        array[j + 1] = key
    
    return array
````

**Explanation:**

The insertionSort function takes an array as input and sorts it in ascending order using Insertion Sort.

The outer loop iterates through the elements of the array starting from the second element (index 1) to the last element. This loop represents the unsorted part of the array.

Inside the loop, key is set to the current element at index i. This element will be "inserted" into its correct position within the sorted part of the array.

j is initialized to i - 1, representing the index of the last element in the sorted part of the array.

The while loop checks if the element at index j is greater than the key and if j is greater than or equal to 0 (ensuring we don't go out of bounds). If both conditions are met, it shifts the element at j to the right (i.e., array[j + 1] = array[j]) to make space for the key.

This shifting process continues until we find the correct position for the key in the sorted part of the array.

Finally, after the while loop, we place the key at its correct position in the sorted part of the array by assigning it to array[j + 1].

The outer loop continues until all elements have been processed and placed in their correct positions.

The sorted array is returned.

**Analysis:**

The time complexity of Insertion Sort is generally O(n^2) in the worst and average cases, where 'n' is the number of elements in the array. This is because, in the worst case, for each element, it may need to make comparisons and potentially shift all the way to the beginning of the sorted part.

However, in the best case, when the array is already nearly sorted, the time complexity can be close to O(n) because there will be fewer comparisons and shifts.

Insertion Sort is efficient for small datasets and has the advantage of being an "online" algorithm, meaning it can sort a list as it receives new elements, unlike other sorting algorithms that require all elements to be present before sorting.

# Merge Sort

This is a divide and conquer algorithm. In this algorithm, we split a list in half and keep splitting the list by 2 until it only has a single element. Then we merge the sorted list. We keep doing this until we get a sorted list with all the elements of the unsorted input list.

Merge Sort is a popular and efficient divide-and-conquer sorting algorithm. It works by recursively dividing the input array into smaller halves, sorting those halves, and then merging them back together into a single sorted array. The key step in Merge Sort is the merging process, where two sorted arrays are combined into a single sorted array.


````python
def mergeSort(nums):
    if len(nums) == 1:
        return nums
    
    mid = (len(nums) - 1) // 2
    
    # Recursively sort the left and right halves
    lst1 = mergeSort(nums[:mid + 1])
    lst2 = mergeSort(nums[mid + 1:])
    
    # Merge the sorted halves
    result = merge(lst1, lst2)
    
    return result

def merge(lst1, lst2):
    lst = []
    i = 0
    j = 0
    
    while i <= len(lst1) - 1 and j <= len(lst2) - 1:
        if lst1[i] < lst2[j]:
            lst.append(lst1[i])
            i += 1
        else:
            lst.append(lst2[j])
            j += 1
    
    # Append remaining elements from lst1 or lst2
    if i > len(lst1) - 1:
        while j <= len(lst2) - 1:
            lst.append(lst2[j])
            j += 1
    else:
        while i <= len(lst1) - 1:
            lst.append(lst1[i])
            i += 1
    
    return lst
````

**Explanation**:

The mergeSort function takes an array nums as input and sorts it using the Merge Sort algorithm.

If the length of the input nums is 1, it is already considered sorted, so it returns nums immediately.

Otherwise, it calculates the middle index mid to divide the array into two halves.

It recursively calls mergeSort on the left and right halves of the array.

Once both halves are sorted, the merge function is called to merge them into a single sorted array.

The merge function takes two sorted lists lst1 and lst2 and merges them into a single sorted list lst.

It uses two pointers i and j to traverse lst1 and lst2, respectively, comparing elements at each step and appending the smaller element to the result lst.

After the while loop, if there are remaining elements in either lst1 or lst2, those elements are appended to the result lst.

The merged and sorted list is returned.

**Analysis**

Merge Sort is an efficient sorting algorithm with a consistent time complexity of O(n log n) in the worst, best, and average cases. This makes it a reliable choice for sorting large datasets. The algorithm achieves this efficiency by dividing the problem into smaller subproblems and then efficiently merging them back together.

# The Linear Search

Linear Search is a simple search algorithm that works by examining each element in a list or array one by one until the target element is found or the end of the list is reached. It's sometimes called a "brute-force" search because it checks every element in the list.

This is the simplest solution to a sequential search problem. I will demonstrate this algorithm using the code below. This code searches for an element in a list. Imagine we have a list of numbers `(List = [1, 2, 3, 9, 7, 4])` and we are to search for a specific number:

````python
def search(list1, n):
    i = 0
    
    while i < len(list1):
        if list1[i] == n:
            return True
        i = i + 1
    return False

list1 = [1, 2, 3, 9, 7, 4]
n = 5

if search(list1, n):
    print("Number found")
else:
    print("Number not found")
````

**Explanation**:

The search function takes two arguments: list1, which is the list to be searched, and n, which is the number you are looking for.

It initializes an index variable i to 0, which represents the current position in the list.

The function enters a while loop that continues as long as i is less than the length of list1.

Inside the loop, it checks if the element at the current position (list1[i]) is equal to the target number n. If they are equal, it returns True, indicating that the number is found.

If the element at the current position is not equal to n, it increments i by 1 to move to the next element in the list.

If the loop finishes without finding n, the function returns False, indicating that the number is not found.

After defining the search function, the code creates a list list1 and a target number n.

It then calls the search function with list1 and n as arguments and checks the returned value.

If search returns True, it prints "Number found." Otherwise, it prints "Number not found."

**Analysis:**

The time complexity of Linear Search is O(n) in the worst-case scenario, where 'n' is the number of elements in the list. This is because, in the worst case, the algorithm may have to check every element in the list before finding the target or determining that it's not in the list.

Linear Search is a straightforward and easy-to-understand search algorithm, but it may not be efficient for large datasets compared to more advanced search algorithms like binary search for sorted lists.

**Best-Case Scenario:**

In the best-case scenario, the element you are searching for is found at the very beginning of the list, or it might be the first element itself. This means that you find the target element in the first comparison.

Time Complexity: O(1)
In this case, the Linear Search algorithm performs exceptionally well because it takes only one comparison to find the element.

**Worst-Case Scenario:**

In the worst-case scenario, the element you are searching for is either not present in the list, or it's located at the very end of the list. This means that you have to examine every element in the list to determine that the element is not there.

Time Complexity: O(n)
In the worst case, you have to check all 'n' elements in the list before concluding that the element is not present. This is the least efficient case for Linear Search.

**Average-Case Scenario:**

In the average-case scenario, the element you are searching for is not found at the beginning or end of the list but could be located anywhere within the list. On average, you might need to examine half of the list before finding the element or determining that it's not present.

Time Complexity: O(n/2) or simply O(n)
On average, Linear Search makes approximately half of the total comparisons before finding the target element or concluding that it's not there.

Summary:

Best Case: O(1) - One comparison is needed to find the element at the very beginning of the list.
Worst Case: O(n) - In the worst case, you need to check all elements in the list.
Average Case: O(n) - On average, you need to check about half of the elements in the list.
