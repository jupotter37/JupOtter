# Big O Notation

The term Big O Notation refers to the analysis of an algorithm's efficiency. Data sets given to an algorithm can affect the time and space complexity in terms of N. Time complexity is the time it takes for the algorithm to run and space complexity is the amount of memory used by an algorithm. This includes the number of inputs in an algorithm. Both of these complexities have best and worst case scenarios of how efficienty the algorithm runs. 

In Big 0, we look at the worst case because developers don't always expect the algorithm to run in the best case, so in the event of a worst case scenario, they analyse which is better. The diagram below shows us how each sorting algorithm at it's worst case performs. 0(n2) performs the worst compared to 0(1) because the more input given to 0(1) doesn't affect the time it takes to computate.

# Turing Machine

A turing machine refers to a way of writing down information in coded form

the turing machine works by divivding a tape up into squares containing a 1 or 0. Each individual square is read one at a time. The squares represent a question or problem we need to solve. These problems can include turning every zero to a one and vice versa. The turing machine will change a square to a different number if it meets the condition and move on to the next square. Once the turing machine has answered the problem, it move to the halting stage and whats left on the tape is the answer to our problem.

Turing machines today are mostly used for research and teaching tools to demonstrate what happens in a CPU.

The following code snippets below demonstrates the idea of the Turing Machine and how the code can be changed to find a specific number like 1.

# Lab Example

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('01110001')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    #for i in range(len(tape)):
    #if tape[i] == 1:
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```

# How Turing Machine relates to quantum computing

The reason turing Machines are related to quantum computing is because it's an abstract model equivalent to the quantum circuit and can represent all features of quantum computer without entanglement. What this means is that the tape used in the turing machine can be replaced with quantum bits or qubits. These qubits are used to represent a point on a chosen square and it's main characteristic of this alternative system is that it permits the coherent superposition of ones and zeros.

This means that the quantum turing machine isn't limited by the constraints of the turing machine like not recognising real numbers and being affected by the halting problem, which is when no program can be written that can predict whether or not any other program halts after a finite number of steps.
