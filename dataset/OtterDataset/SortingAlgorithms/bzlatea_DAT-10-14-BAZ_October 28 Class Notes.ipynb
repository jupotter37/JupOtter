```python
conditions = [   
    (df['Survived'] == 1) & (df['Sex'] == 'female'),
    (df['Survived'] == 1) & (df['Sex'] == 'male'),
    (df['Survived'] == 0) & (df['Sex'] == 'female'),
    (df['Survived'] == 0) & (df['Sex'] == 'male')
]

results = [
    'Female-Survived',
    'Male-Survived',
    'Female-Died',
    'Male-Died'
]


df['Status'] = np.select(conditions, results,'Other')
```

```python
def over_75(val):
    if val > 75:
        return 'Getting up there'
    else:
        return 'In the Bloom of the youth'
```

```python
df['Demographic'] = df['Age'].apply(over_75)
```

```python
%timeit df['Demographic'] = df['Age'].apply(over_75)
```

```python
def get_status(val):
    if val['Survived'] == 1 and val['Sex'] == 'female':
        return 'Female-Survived'
    elif val['Survived'] == 1 and val['Sex'] == 'male':
        return 'Male-Survived'
    elif val['Survived'] == 0 and val['Sex'] == 'female':
        return 'Female-Died'
    elif val['Survived'] == 0 and val['Sex'] == 'male':
        return 'Male-Died'
    else:
        'Other'
```

```python
#now looking up across different columns
#so what we do, is apply it to the whole dataframe and SPECIFY the axis that we will run across the row
df['Status'] = df.apply(get_status, axis = 1)
```

```python
df['Is_Alone'] = np.where(df['Family Size']>0, True, False)
```

```python
conditions = [
    (df['Sex'] == 'female') & (df['Pclass'] == 1),
    (df['Sex'] == 'female') & (df['Pclass'] > 1),
    (df['Sex'] == 'male') & (df['Pclass'] == 1),
    (df['Sex'] == 'male') & (df['Pclass'] > 1),
    
]

results = [
    'F-High',
    'F-Low',
    'M-High',
    'M-Low'
]
```

```python
df['Gender_Status'] = np.select(conditions, results,'Other')
```
