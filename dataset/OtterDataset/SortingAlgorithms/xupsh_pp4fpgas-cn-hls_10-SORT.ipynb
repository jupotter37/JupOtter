# Insertion sort is a basic sorting algorithm whose core idea is to insert a new element into an ordered array and keep it in order. In each step, insert a record to be sorted into the appropriate place in the previously sorted file according to the size of its key value until all the records are inserted.
插入排序是一种基本的排序算法，其核心思想是将一个新的元素插入到一个有序数组中，并继续保持有序。每步将一个待排序的记录，按其关键码值的大小插入前面已经排序的文件中适当位置上，直到全部插入完为止。

```python
from pynq import Xlnk

xlnk = Xlnk()
in_buffer = xlnk.cma_array(shape=(32,), dtype=np.int) #the length of the 
#array has to be twice the number you wanna sort
#here we set the sorted array's length to be 16
out_buffer = xlnk.cma_array(shape=(16,), dtype=np.int)

for i in range(16):
    in_buffer[i] = random.randint(0,100);

dma.sendchannel.transfer(in_buffer)
# dma.sendchannel.wait()
dma.recvchannel.transfer(out_buffer)

# dma.recvchannel.wait()


actualin = in_buffer[0:16]
    
print("input is:")   
print(actualin)
print("sorted output:")
print(out_buffer)
```
