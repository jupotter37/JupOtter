```python
import timeit
import random

#Implementing the 3 sorting algorithms:

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        index = i

        for j in range(i + 1, n):
            if arr[j] < arr[index]:
                index = j
        arr[i], arr[index] = arr[index], arr[i]

    return arr

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr



```

```python
# Function to get a list of tuples containing number of elements in an array and time taken to sort that array using the Algorithms

def get_benchmarks():
    benchmarks = [[0], [0], [0]]
    for i in range(1, 1001):
        length = i
        arr = [random.randint(-1000, 1000) for _ in range(length)]
        bubble_time = timeit.timeit(lambda: bubble_sort(arr.copy()), number=1) * 1000000
        insertion_time = timeit.timeit(lambda: insertion_sort(arr.copy()), number=1) * 1000000
        selection_time = timeit.timeit(lambda: selection_sort(arr.copy()), number=1) * 1000000
        benchmarks[0].append(bubble_time)
        benchmarks[1].append(insertion_time)
        benchmarks[2].append(selection_time)
    return benchmarks
```

```python
benchmarks = get_benchmarks()
```

```python
import matplotlib.pyplot as plt

sizes = list(range(1, 1002)) 

bubble_times = benchmarks[0]  
insertion_times = benchmarks[1]  
selection_times = benchmarks[2]  

# Plotting the data
plt.figure(figsize=(10, 6))

plt.plot(sizes, bubble_times, label='Bubble Sort', color='b', marker='o')
plt.plot(sizes, insertion_times, label='Insertion Sort', color='r', marker='o')
plt.plot(sizes, selection_times, label='Selection Sort', color='g', marker='o')

plt.xlabel('Array Size')
plt.ylabel('Time Taken (s)')
plt.title('Sorting Algorithm Performance')
plt.legend()
plt.grid(True)
plt.show()
```
