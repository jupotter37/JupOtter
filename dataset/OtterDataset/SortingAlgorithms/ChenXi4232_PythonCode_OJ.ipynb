```python
class DisjointSet:
    def __init__(self, size):
        self.p = [i for i in range(size)]
        self.r = [0] * size

    def find(self, x):
        if self.p[x] != x:
            self.p[x] = self.find(self.p[x])
        return self.p[x]

    def union(self, x, y):
        r_x = self.find(x)
        r_y = self.find(y)
        if r_x != r_y:
            if self.r[r_x] < self.r[r_y]:
                self.p[r_x] = r_y
            elif self.r[r_x] > self.r[r_y]:
                self.p[r_y] = r_x
            else:
                self.p[r_x] = r_y
                self.r[r_y] += 1


class Edge:
    def __init__(self, x, y, length):
        self.x = x
        self.y = y
        self.length = length


def main():
    n, m = list(map(int, input().split()))
    # print(union)
    disjoint_set = DisjointSet(n)
    G_edge = []
    for i in range(m):
        x, y, length = list(map(int, input().split()))
        G_edge.append(Edge(x, y, length))
    G_edge.sort(key=lambda G_edge: G_edge.length)
    # print(G_edge[0].x)
    Tree_sum = 0
    i = 0
    j = 0
    while i < n-1 and j < m:
        if disjoint_set.find(G_edge[j].x-1) != disjoint_set.find(G_edge[j].y-1):
            disjoint_set.union(G_edge[j].x-1, G_edge[j].y-1)
            Tree_sum += G_edge[j].length
            j += 1
            i += 1
        else:
            j += 1
    if i != n-1:
        return -1
    return Tree_sum


if __name__ == '__main__':
    print(main())
```

```python
from collections import deque

# 读取输入
n, m = map(int, input().split())
adj = [[] for _ in range(n+1)]
for i in range(m):
    x, y = map(int, input().split())
    adj[x].append(y)
    adj[y].append(x)

# 初始化距离和路径条数
dist = [-1] * (n+1)
paths = [0] * (n+1)
dist[1] = 0
paths[1] = 1

# BFS遍历
queue = deque([1])
while queue:
    u = queue.popleft()
    for v in adj[u]:
        if dist[v] == -1:  # 如果v未被访问过
            dist[v] = dist[u] + 1
            paths[v] = paths[u]
            queue.append(v)
        elif dist[v] == dist[u] + 1:  # 如果v已被访问过，且是最短路径
            paths[v] += paths[u]

# 输出结果
for i in range(2, n+1):
    print(paths[i] if dist[i] != -1 else 0)
```

```python
from collections import deque

MOD = int(1e9) + 7

num_people, num_relationships = map(int, input().split())

adjacency_list = [[] for _ in range(num_people+1)]
reverse_adjacency_list = [[] for _ in range(num_people+1)]
in_degree = [0] * (num_people+1)
out_degree = [0] * (num_people+1)

for i in range(num_relationships):
    person_a, person_b = map(int, input().split())
    adjacency_list[person_a].append(person_b)
    reverse_adjacency_list[person_b].append(person_a)
    in_degree[person_b] += 1
    out_degree[person_a] += 1

queue = deque([i for i in range(1, num_people+1) if out_degree[i] == 0])
num_paths = [1 if out_degree[i] == 0 else 0 for i in range(num_people+1)]
answer = 0

while queue:
    vertex = queue.popleft()
    for neighbor in reverse_adjacency_list[vertex]:
        num_paths[neighbor] += num_paths[vertex]
        num_paths[neighbor] %= MOD
        out_degree[neighbor] -= 1
        if out_degree[neighbor] == 0:
            queue.append(neighbor)

for i in range(1, num_people+1):
    if in_degree[i] == 0:
        answer += num_paths[i]
        answer %= MOD

print(answer)
```

```python
import math


def count_primes(start_num, end_num):
    if start_num > end_num:
        return 0

    is_prime = [True] * (end_num + 1)
    is_prime[0] = is_prime[1] = False

    sqrt_end = int(math.sqrt(end_num)) + 1
    for num in range(2, sqrt_end):
        if is_prime[num]:
            for multiple in range(num*num, end_num+1, num):
                is_prime[multiple] = False

    prime_count = sum(is_prime[start_num:end_num+1])
    return prime_count


start_num, end_num = map(int, input().split())

prime_count = count_primes(start_num, end_num)
print(prime_count)
```
