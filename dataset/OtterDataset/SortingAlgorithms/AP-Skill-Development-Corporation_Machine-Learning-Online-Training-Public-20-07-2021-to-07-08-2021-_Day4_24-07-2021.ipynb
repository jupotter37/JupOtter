# Handling Missing Data

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/titanic.csv")
titanic.head()
```

```python
titanic.shape
```

```python
# finding missing values

titanic.isnull().sum()
```

```python
titanic.notnull().sum()
```

```python
titanic.columns
```

```python
titanic.dropna().shape # drops the data having any one missing value in the record
```

```python
891-183
```

```python
titanic.dropna()
```

```python
titanic.drop("sibsp", axis = 1 ) # drop a particular column
```

```python
titanic.drop_duplicates()
```

```python
titanic.duplicated().sum() # finding duplicated data in the datset
```

```python
titanic.isnull().sum()
```

```python
titanic["embarked"].value_counts()
```

```python
titanic[titanic["embarked"].isnull() == True]["embarked"] = titanic["embarked"].mode()
```

```python
644+168+77  
891-889
```

```python
titanic["embarked"].isnull().sum()
```

```python
titanic[titanic["embarked"].isnull() == True]
```

```python
titanic.loc[[61,829],"embarked"] = "S"
```

```python
titanic["embarked"].isnull().sum() # filling missing value
```

```python
titanic["cabin"].isnull().sum()
```

```python
titanic["cabin"].fillna(method= "pad") # filling the value with previous one
```

```python
titanic["cabin"].fillna(method= "pad").isnull().sum()
# fillna means it replace only null value into given data
```

```python
titanic["cabin"].fillna(method= "bfill", inplace = True) # filling with next one/ backword value
```

```python
titanic.isnull().sum()
```

```python
help(titanic["age"].replace)
# is use to replace the value by given value
```

```python
titanic["age"].replace(to_replace=np.nan, value = -1, inplace = True )
```

```python
titanic.isnull().sum()
```

```python
titanic[titanic["cabin"].isnull() == True]
```

```python
titanic["cabin"].loc[890] = "C50"
```
