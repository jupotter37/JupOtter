```python
from scipy import stats 
```

```python
data = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
data['start_ts'] = pd.to_datetime(data['start_ts'])

sabados_chuvosos = data[data['weather_conditions'] == 'Bad']['duration_seconds']
sabados_nao_chuvosos = data[data['weather_conditions'] == 'Good']['duration_seconds']

_, p_value_chuvosos = stats.shabpiro(sabados_chuvosos)
_, p_value_nao_chuvosos = stats.shapiro(sabados_nao_chuvosos)
_, p_value_variancias = stats.levene(sabados_chuvosos, sabados_nao_chuvosos)

t_statistic, p_value_ttest = stats.ttest_ind(sabados_chuvosos, sabados_nao_chuvosos, equal_var=True)

alfa = 0.05

if p_value_ttest < alfa:
    print(f'Rejeitar a hipótese nula. P-value: {p_value_ttest:.4f}')
else:
    print(f'Não rejeitar a hipótese nula. P-value: {p_value_ttest:.4f}')
```

A duração média dos passeios do Loop para o Aeroporto Internacional O'Hare parece variar significativamente entre sábados chuvosos e sábados não chuvosos.
