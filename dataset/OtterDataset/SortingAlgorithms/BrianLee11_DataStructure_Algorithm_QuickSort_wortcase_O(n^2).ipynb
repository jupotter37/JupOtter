#**Demonstrate why the time complexity for quick sort in the worst case is $O(n^2)$**

**QuickSort** is one of the best sorting algorithms developed by Tony Hoare in 1960 to sort the array. It follows the **divide-and-conquer** rule like Merge Sort but unlike **Merge Sort**, this algorithm does not use any extra space for sorting (though it uses an auxiliary stack space).

The basic idea behind **QuickSort** is to select a “pivot” element from the array and partition the other elements into two sub-arrays according to whether they are less than or greater than the pivot.

Worst-case in quicksort occurs when the pivot selection results in the most unbalanced partitions possible, such as when the smallest or largest element is consistently chosen as the pivot. In such cases, the partitioning step does not effectively divide the array, leading to suboptimal performance.

The worst-case recurrence relation for Quicksort is:

$$T(n) = T(n-1) + O(n)$$

**Explanation:**
- T(n−1): This represents the time complexity of sorting the n−1 elements that remain after the pivot element is chosen and the partitioning is done.
- O(n): This represents the time complexity of partitioning the array around the pivot element, which takes linear time.


**Expanding the Recurrence:**
We expand the recurrence relation by repeatedly substituting $T(n-1)$ :
$$
\begin{aligned}
T(n) & =T(n-1)+O(n) \\
& =(T(n-2)+O(n-1))+O(n) \\
& =((T(n-3)+O(n-2))+O(n-1))+O(n) \\
& \vdots \\
& =T(1)+\sum_{k=2}^n O(k)
\end{aligned}
$$

**Summation:**
The summation $\sum_{k=2}^n O(k)$ can be approximated as:
$$
\sum_{k=2}^n O(k) \approx \sum_{k=1}^n k=1+2+3+\ldots+n
$$

This is the sum of the first $n$ natural numbers, which is given by the formula:
$$
\sum_{k=1}^n k=\frac{n(n+1)}{2}
$$

**Time Complexity:**
The formula $\frac{n(n+1)}{2}$ is a quadratic expression and can be simplified to $O\left(n^2\right)$ :
$$
\frac{n(n+1)}{2} \in O\left(n^2\right)
$$
