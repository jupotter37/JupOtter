```python
#Wap to display sorted array using merge sort and number of iterations also

def mergesort(array):
  if len(array) > 1:
    r = len(array)//2
    L = array[:r]
    M = array[r:]

    mergesort(L)
    mergesort(M)

    i = j = k = t = 0

    while i < len(L) and j < len(M):
      if L[i] < M[i]:
        array[k] = L[i]
        i += 1
        t += 1
      else:
          array[k] = M[j]
          j += 1
      k += 1

    while i < len(L):
      array[k] = L[i]
      i += 1
      k += 1
    while j < len(M):
      j += 1
      k += 1

def printlist(array):
  for i in range(len(array)):
    print(array[i], end = " ")
  print()

if __name__ == "__main__":
  L1= [34,56,12,78,2,33,78]
  mergesort(L1)
  print("Sorted array is", printlist(L1))
```
