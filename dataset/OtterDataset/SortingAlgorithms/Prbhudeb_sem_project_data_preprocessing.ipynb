```python
dt.rename(columns={'Skills': 'Skills Required', 'Complexity': 'Difficulty Level'}, inplace=True)

# Function to assign projects
def assign_project(row, data):
    matching_projects = data[
        (data['Skills Required'].str.contains(row['Skills Required'], na=False)) &
        (data['Difficulty Level'] == row['Difficulty Level'])
    ]
    if not matching_projects.empty:
        return matching_projects.iloc[0]['Project Name']
    return "No suitable project"

# Apply project assignment
dt['Recommended_Project'] = dt.apply(
    lambda row: assign_project(row, data), axis=1
)
```
