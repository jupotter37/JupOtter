Con las correcciones en los archivos CSV, 
hay que rehacer lo que se hizo en la segunda práctica para actualizar los archivos:

```python
import pandas as pd
enc_ini = pd.read_csv("encuestas/ini.csv")
enc_mcu = pd.read_csv("encuestas/mc.csv")
enc_ord = pd.read_csv("encuestas/ord.csv")
cf = pd.read_csv("profes.csv", header=None)
enc_mcu['matr'] = [str(valor) for valor in enc_mcu['matr']]
cf.columns = ['profe', 'sem', 'matr', '1ra', '2da']
cf['matr'] = [str(valor) for valor in cf['matr']]
enc_ini['profe'] = ['elisa' if 'Elisa' in str(valor) else 'moi' if 'Moi' in str(valor) else 'NA' for valor in enc_ini['grupo']]
enc_mcu['profe'] = ['elisa' if 'Elisa' in str(valor) else 'moi' if 'Moi' in str(valor) else 'NA' for valor in enc_mcu['grupo']]
enc_ord['profe'] = ['elisa' if 'Elisa' in str(valor) else 'moi' if 'Moi' in str(valor) else 'NA' for valor in enc_ord['grupo']]
tmp = pd.merge(enc_ini, enc_mcu, how='outer', on=['profe', 'sem', 'matr'])
tmp2 = pd.merge(tmp, enc_ord, how='outer', on=['profe', 'sem', 'matr'])
d = pd.merge(tmp2, cf, how='outer', on=['profe', 'sem', 'matr'])
d.to_csv('datos.csv')
casos = []
for i in d.index:
    cf1ra = str(d['1ra'][i])
    cf2da = str(d['2da'][i])
    if cf1ra.isdigit() and int(cf1ra) >= 70:
        casos.append('1ra')
    elif cf2da.isdigit() and int(cf2da) >= 70:
        casos.append('2da')
    elif cf1ra.isdigit() and cf2da.isdigit():
        casos.append('3ra')
    else:
        casos.append('NA')
d['caso'] = casos
d.to_csv('casos.csv')
```

```python
>>> valores = {'70-79': 75, '80-89': 85, '90-100': 95, 'segunda': 50}
>>> d.califEsp = d.califEsp.replace("Creo que aprobaré en segunda oportunidad", "segunda")
>>> d.califEsp_ini = d.califEsp_ini.replace("Estimo aprobar en segunda oportunidad", "segunda")
>>> d.califEsp_mcu = d.califEsp_mcu.replace("Estimo aprobar en segunda oportunidad", "segunda")
>>> d['eI'] = [valores.get(v, NaN) for v in d.califEsp_ini]
>>> d['eM'] = [valores.get(v, NaN) for v in d.califEsp_mcu]
>>> d['eF'] = [valores.get(v, NaN) for v in d.califEsp]
>>> d.CF1op.corr(d.eI)
0.1957813659194152
>>> d.CF1op.corr(d.eM)
0.4724877223145954
>>> d.CF1op.corr(d.eF)
0.45281735470132939
```
