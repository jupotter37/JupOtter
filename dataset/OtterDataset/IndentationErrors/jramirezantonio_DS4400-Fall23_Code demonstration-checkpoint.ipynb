#### mtcars data: EDA and fit the models

```python
import statsmodels.api as sm
data = sm.datasets.get_rdataset('mtcars').data 
data.head(6)
```

```python
import matplotlib.pyplot as plt

plt.hist(data.mpg, bins = 20)
plt.xlabel("mpg")
plt.title("The histogram of mpg")
```

```python
plt.scatter(data[['wt']], data[['mpg']])
plt.ylabel('miles per gallon')
plt.xlabel('weight')
plt.title('The scatter plot between mpg and wt')
```

```python
import seaborn as sns

sns.lmplot(x = 'wt', y = 'mpg', data = data, aspect = 2)
plt.title('Scatter plot with linear regression line for wt and mpg')
plt.show()
```

```python
import sklearn
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

X = data.drop('mpg', axis=1)
Y = data.mpg
print(X)
print(Y)
```

```python
lm.fit(X[['wt']], Y)

print("The estimated intercept coefficient is %.3f" %lm.intercept_)
print("the estimated coefficient for weight is %.3f" %lm.coef_)
```

```python
lm.fit(X, Y)
print(lm.intercept_)
print(lm.coef_)
```

```python
import pandas as pd
from pandas import DataFrame

lm.fit(X, Y)
coef_df = DataFrame(X.columns)
# print(coef_df)
coef_df.columns  = ["Feature"]
# print(coef_df, coef_df.columns)
coef_df['Coefficient Estimate'] = pd.Series(lm.coef_)
print(coef_df)

print("The intercept coefficient is %.3f" %lm.intercept_)
```

#### Interpretation of linear model

One of the advantages of linear regression is tnat it is one of the models with the easiest interpretation. In the area of machine learning, this may not be important but if you are interested in the relationship between the target and features, linear regression will be super helpful. 

Recall our first regression with only "wt" as the feature. We have the estimated regression as: 

$$ h_{\theta}(X) = 37.285 - 5.344 X $$

- The interpretation for the intercept means when $X = 0$, the estimated target value, even though our $X$ may not be 0 in the practice. 
- The interpretation for the slope means with one unit increases on $X$ value, the change of $h_{\theta}(X)$. 

In this example, when weight of the car is 0 (ton), the estimated mpg is 37.285 (even though this is impossible). For one unit (ton) increase on the weight of the car, the mpg will drop 5.344. 

The interpretation of multiple linear regression is almost the same, except when we interpret each coefficient, for example, disp(displacement)'s coefficient is 0.013335, we will say: When all the other features hold the same
