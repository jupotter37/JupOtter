```python
#1
# input 값으로  "name, 학번, 국, 영, 수, 과학 성적"을 입력받아 다음 출력값을 갖도록 하는 함수(함수이름: score)를 만들어라:
# (1) "score('홍길동', 2020030, 100, 80,90, 70)"를 실행했을 때, 다음 출력결과가 나오도록 하고
# 출력결과: ***학생(학번)은 평균 ***점입니다.
# (2) 평균값을 return하도록 하여라.
name = str(input('이름:'))
h = int(input('학번:'))
args = (int(input('국어')), int(input('수학')), int(input('영어')), int(input('과학')))

def score(name, h, *args):
    average = sum(args) / len(args)
    print(f'{name}학생 {학번}은 평균 {average:.2f}점입니다.')

score(name, h, *args)
```
