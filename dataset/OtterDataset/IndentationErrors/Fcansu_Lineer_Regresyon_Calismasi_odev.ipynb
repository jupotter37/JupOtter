```python
cansunun_verisi["oran"]= cansunun_verisi["kisibasigelir"]/cansunun_verisi["saglikharcamasi"]
cansunun_verisi.head(14)
```

```python
plt.scatter(cansunun_verisi["oran"], cansunun_verisi["kisibasigelir"], color="red")
plt.xlabel("Oran")
plt.ylabel("Kişi Başı Gelir")
plt.title("Gelir-Sağlık harcamaları Oranı Grafiği")
plt.show()
```

```python
a = 0.001
b = 0.8
def oran_tahmin(kisibasigelir):
   return a * kisibasigelir + b
```

```python
asgariucret_tahmin = oran_tahmin(17500)
print(asgariucret_tahmin)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
```

```python
model.fit([[7500],[8000],[11000]],[0.0002,0.0023,0.5])
```

```python
model.fit(cansunun_verisi[["kisibasigelir"]],cansunun_verisi["oran"])
```

```python
tahmin_edilen_degerler= model.predict(cansunun_verisi[["oran"]])
tahmin_edilen_degerler
```

```python
model.coef_,model.intercept_
```

```python
model.predict([[1000]])
```

```python
plt.scatter(cansunun_verisi["oran"],cansunun_verisi["kisibasigelir"],color="red")
plt.plot(cansunun_verisi["oran"],tahmin_edilen_degerler, color="blue")
plt.xlabel("Kişi Başı Gelir")
plt.ylabel("Oran")
plt.title("Gelir-Sağlık harcamaları Oranı Grafiği")
plt.show()
```

```python
model.score(cansunun_verisi[["oran"]], cansunun_verisi["kisibasigelir"])
```
