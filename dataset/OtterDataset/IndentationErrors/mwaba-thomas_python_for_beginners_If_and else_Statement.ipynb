##### ASKING QUESTIONS WITH IF AND ELSE

###### In programming, we often ask yes or no questions and choose what do something based on the answer;
###### for example "How old are you?" and if the answer is "greater than 1", answer back with " you are not a baby"                                            These questions are called *conditions*

##### An if statement is written like this in python

```python
age = 8
if age > 1:
    print("you are not a baby")
```

##### A BLOCK IS A GROUP OF PROGRAMMING STATEMENTS
###### You might want to do more than print "You are not a baby" maybe you would like to print out other sententces.

```python
age = 14
if age > 1:
    print("you are not a baby")
    print("Why are you crawling?")
    print("Did you forget how to walk?")
```

###### This block code is made up of three print statements that run only if age > 1 is true.

###### A block with four spaces on one line and six spaces on the next will give an indentation error when you run it.
###### Indentation is like giving your code a clean and tidy appearance by adding spaces or tabs at the beginning of lines. It makes things neat and easy to follow.

```python
age = 14
if age > 1:
    print("you are not a baby")
      print("Why are you crawling?")
```

##### IF - THEN - ELSE STATEMENTS

###### We use an if-then-else statement, which says  “If something is true, then do this; or else, do that.”

```python
age_ = 18
if age_ >= 18:
    print("You can cook for yourself")
else:
    print("Wait for your parents")
```
