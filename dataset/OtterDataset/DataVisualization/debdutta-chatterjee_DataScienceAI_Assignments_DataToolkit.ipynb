###7. What is a heatmap, and when should it be used?

Heatmap is a colour coded matrix to visualize data.
It can be used to visualize correlation between data or to analyze patterns over time. It can be used to depict intensity of data.

Example - A football players presence inside the different area of the pitch can be visualized using heatmap

```

import seaborn as sns

#sns.get_dataset_names()

df_numeric.corr()

plt.title("Tips data set heatmap",fontweight="bold",fontsize=15)
plt.show()
```

###8. What does the term “vectorized operation” mean in NumPy?

Vectorized operations work on the entire array using instead of element wise operation. The operations are done in a single statement without using loops.

```

print(arr1)
print(arr2)

arr3  = arr1*arr2
print(arr3)
```

````python
###8. Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dates = pd.date_range(start="2010-01-01",end="2025-01-01",freq="D")
sales_date = np.random.choice(dates,size=10,replace=False)

sales_car = np.arange(1000,21000,2000)
sales_bike = np.arange(1000,11000,1000)
sales_house = np.arange(3000,13000,1000)

plt.plot(sorted(sales_date),sales_car,label="Car sales",marker="X")
plt.plot(sorted(sales_date),sales_bike,label="Bike sales",marker="o")
plt.plot(sorted(sales_date),sales_house,label="House sales",marker="*")
plt.grid()
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.title("Sales Data",fontsize=10,fontweight="bold")
plt.show()

````

````python
###11. Perform matrix multiplication using NumPy

import numpy as np

mat1 = np.random.randint(1,10,size=(2,3))
mat2 = np.random.randint(1,10,size=(3,2))

mult = mat1 @ mat2
print(mult)
print()

#Alternate way
mult = np.matmul(mat1,mat2)
print(mult)
````
