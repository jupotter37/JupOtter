**데이터 시각화 (Data visualization)의 규칙**

## **범용적인 규칙**

```python
# 데이터 시각화의 기본
1. 시각적 속성은 위치,형태,크기,색,선 굵기,선 유형 을 말하며
    위치,크기,색,선 굵기는 연속적 유형(float)에서 사용되지만
    형태(동그라미,네모,세모), 선 유형(실선, 점선)은 이산형 데이터만을 표현한다.

2. 데이터 값마다 단 하나의 시각적 속성 값이 대응되어야 한다.
```

```python
# 그래프의 축과 스케일
3. X,Y축이 동일한 단위로 그려져 있으면 그래프의 가로 세로 비율을 동일하게(격자를 정사각형으로)
    만들어야 하며, 그렇지 않다면 데이터를 잘 표현할 수 있도록 비율을 다르게 해야한다.

4. 비율처럼 곱셈이나 나눗셈으로 얻은 값은 로그 스케일로 표시하면 자연스럽게 표시할 수 있으며
    데이터셋의 값이 들쭉날쭉할 때도 로그 스케일을 많이 사용한다.

5. 제곱근(root) 스케일은 적절한 위치에 눈금을 표기하는 것이 까다롭지만 제곱으로 얻은 데이터(땅의 면적)를
    그래프로 표기하기에 적합하다.

6. 극 좌표계(polar coordinate system)는 축이 원형을 이루는 좌표계로 특정 기간에 생성되는 데이터를
    표기하는데 유용하다(1년 동안의 기온을 극 좌표계에 그리면 계절별 온도 편차가 큰 지역과 작은 지역을
    한 눈에 파악할 수 있다).
```

```python
# 색상 스케일
7. 데이터 시각화에서 색을 사용하는 경우는 세 가지다.
    1) 데이터 군을 서로 구분하기 위해
        정성적 색상 스케일(서로 다른 색상)을 사용한다. 한정된 색을 사용해 구분하며,
        나란히 있을때 유난히 튀는 색이 있거나 색상으로 순서가 정해져 있다는 인상을 줘서는 안된다.
        (Okabe Ito, ColorBrewer Dark2, ggplot2 hue 등의 색상 스케일이 있다.)

    2) 데이터 값을 나타내기 위해
        순차적 색상 스케일(진해지는 색상)을 사용한다. 기온, 속력, 소득과 같은
        정량적 데이터 값을 색으로 표현하기 위해 사용하며 색상 스케일을 이루는 색이 전체적으로
        균일하게 차이를 보여야 한다. 연한 하늘색부터 짙은 남색까지 하나의 색조로 표시하는 것을
        단색상 스케일(ColorBrewer Blues)이라고 하고, 여러 색조로 이루어진 것을 다색상 스케일(Viridis)이라고 한다.
        경우에 따라서는 0 혹은 50%를 기준으로 양수, 음수를 색상으로 구분해야 할때도 있는데 이때 사용하는
        색상 스케일을 발산형 색상 스케일(CARTO Eath, Blue-Red)이라고 한다.

    3) 데이터 값을 강조하기 위해
        색은 데이터의 특성 요소를 강조하기에 효과적인 요소이다. 내가 원하는 부분을 다른 부분과
        확실히 구분되는 강렬하고 선명한 색으로 사용하면 전달하고자 하는 내용을 더욱 쉽게 전달할 수 있다.
        이때 사용하는 색상 스케일을 강조 색상 스케일(Okabe Ito, ColorBrewer)이라고 한다.
        강조 스케일을 사용할때는 강조하고자 하는 부분이 더욱 두드러지게 하기 위해서
        다른 부분에 적용하는 기본 색을 은은하고 밋밋한 색상으로 해야한다.(나머지의 색을 없애는 것도 하나의 방법이다)
```

```python
# 다양한 시각화 방식
8. 데이터를 시각화할 때 적절한 그래프와 차트를 사용하면 더욱 명확하게 내용을 전달할 수 있다.
    1) 수량의 시각화
        - 막대 그래프(묶은 막대, 누적 막대), 산점도 그래프, 히트맵(색으로 수량을 표시)

    2) 분포의 시각화
        - 히스토그램, 밀도 도표 : 분포 데이터를 가장 직관적으로 보여주지만 파라미터를 임의로 정해야 해서 유의해야 한다.
        - 누적 밀도 도표, QQ 도표 : 항상 데이터를 정확하게 보여주지만 의미를 파악하기가 까다롭다.
        - 박스플롯, 바이올린 도표, 스트립 차트, 시나 도표 : 여러 분포 값을 한 번에 나타내거나,
                                                            전반적인 변화 추이를 볼때 유용하다.
        - 누적 히스토그램, 중첩 밀도 도표 : 분포 값의 개수가 적을 때 비교 분석을 하기에 좋지만,
                                            누적 히스토그램은 피하는 것이 낫다.
        - 융기선 도표 : 바이올린 도표 대신 쓰기 좋으며 분포 값이 많거나, 시간의 흐름에 따라 분포가 변화할때 유용하다,

    3) 비율의 시각화
        - 파이 차트 : 파이 차트는 각 데이터의 비율을 한 눈에 알아보기 쉽게 한다. 단일 데이터셋에 유용하다.
        - 막대 그래프(묶은 막대, 누적 막대) : 단일 데이터 셋에서 막대 그래프는 각 데이터의 비율을 쉽게 비교할 수 있게 하고,
                                              누적 막대 그래프나 묶은 막대 그래프는 여러 개의 데이터셋을 다룰때 효과적이다.
        - 누적 밀도 도표 : 연속형 변수를 따라 비율이 달라지는 데이터를 나타내기에 좋다.

    4) X-y 관계로 나타내는 시각화
        - 산점도 : 하나의 정량적 변수를 다른 변수와 비교해 나타낼 때 쓰는 가장 대표적인 방식으로
                   변수가 3개일때는 크기를 추가한 버블 차트로 만들어 사용한다.
        - 밀도 등고선 : 데이터 포인트가 매우 많은 경우 산점도로 나타내면 점들이 겹쳐서 보기 어려우므로 이때는
                        등고선이나 2차원 상자, 육각형 상자를 사용하면 좋다.
        - 상관 곡선 : 시각화하려는 정량적 데이터가 두 가지 이상이라면,
                      상관곡선(correlogram)의 형태의 상관계수로 나타낼 수 있다.
        - 꺾은선 그래프 : x축의 데이터가 시간, 의약품 투여량처럼 엄밀하게 증가하는 수치인 경우에는
                          주로 꺾은선 그래프를 그린다.

    5) 지리공간 데이터의 시각화
        - 단계구분도 : 지리공간 데이터는 기본적으로 지도(map)위에 표시하며, 데이터를 기준으로 각기 다른 색을 칠해서
                       지역별 데이터 값을 시각화할 수 있다.

    6) 불확실성의 시각화
        - 오차 막대(error bar) : 추정하거나 측정한 데이터의 예상 값 범위를 나타내며,
                                 예상 값 기준치에서 가로나 세로 방향으로 확장하여 표시한다.
                                 그 기준점은 점, 막대 등으로 표시할 수 있으며, 예상 범위의 신뢰도가
                                 다른점을 반영하여(보통 측정치와 가까울수록 높음) 단계별 오차 막대를 그릴수 있다.
        - (감은)눈 모양 도표 : 융기선 도표와 바이올린 도표의 신뢰 수준과 불확실성 분포 상태를 동시에 보여준다.
        - 신뢰 대역(confidence band) : 곡선 그래프에서의 오차 막대 역할이며 주어진 신뢰 수준 안에서 선이 지나갈 수 있는
                                       값의 범위를 나타내고, 오차 막대와 마찬가지로 단계별 신뢰 대역 도표를 그릴 수 있다.
```

```python
#통합 시각화 함수 목록

1. 수량의 시각화:
Matplotlib: plt.bar (막대 그래프), plt.scatter (산점도), plt.pcolormesh (히트맵)
Seaborn: sns.barplot (막대 그래프), sns.scatterplot (산점도), sns.heatmap (히트맵)
Plotly: plotly.express.bar (막대 그래프), plotly.express.scatter (산점도), plotly.express.imshow (히트맵)
Pandas: pandas.plotting.bar (누적 막대 그래프)

2. 분포의 시각화:
Matplotlib: plt.hist(cumulative=True) (누적 히스토그램), plt.boxplot (박스플롯)
Seaborn: sns.kdeplot (밀도 도표), sns.violinplot (바이올린 도표), sns.stripplot (스트립 차트)
Plotly: plotly.express.histogram (히스토그램), plotly.express.violin (바이올린 도표)
Pandas: pandas.plotting.boxplot (박스플롯)

3. 비율의 시각화:
Matplotlib: plt.pie (파이 차트), plt.bar (막대 그래프)
Seaborn: sns.barplot (막대 그래프)
Plotly: plotly.express.pie (파이 차트), plotly.express.bar (막대 그래프)
Pandas: pandas.plotting.pie (파이 차트)
Statsmodels:  statsmodels.graphics.mosaicplot.mosaic (모자이크 도표)
              statsmodels.graphics.gofplots.qqplot (QQ 도표)
Pandas: pandas.plotting.parallel_coordinates (평행 집합 도표)

4. X-Y 관계로 나타내는 시각화:
Matplotlib: plt.scatter (산점도), plt.plot (꺾은선 그래프), plt.contour (등고선)
Seaborn: sns.scatterplot (산점도), sns.lineplot (꺾은선 그래프),
         sns.kdeplot (등고선 / 밀도 도표), sns.heatmap (히트맵), sns.pairplot (상관곡선)
Plotly: plotly.express.scatter (산점도), plotly.express.line (꺾은선 그래프), plotly.express.density_contour (등고선)
statsmodels = statsmodel.api.nonparametric.lowess(국소 추정 가중 회귀)

5. 지리공간 데이터의 시각화:
Matplotlib: plt.imshow (단계구분도)
Plotly: plotly.express.choropleth (단계구분도)

6. 불확실성의 시각화:
Matplotlib: plt.errorbar (오차 막대)
Seaborn: sns.lineplot (신뢰 대역)
Plotly: plotly.express.scatter (오차 막대 포함 가능)
```

## **다양한 시각화 방법의 활용 방법**

```python
# 1. 수량 데이터의 시각화
9. 막대 도표를 활용해 수치의 크기를 표현할 때, 이름 레이블의 길이가 길어서 가로로 자리를 많이 차지하는 경우
   레이블을 기울여 표기할 수 있지만, 가로 막대 도표로 바꾸면 더욱 읽기 편하게 쓸 수 있다.

10. 기본적으로 범주에 순서가 없다면 오름차순이나 내림차순으로 정렬하여 표기해야 한다.
    그러나 연령별 구분처럼 순서가 있는 경우에는 그 순서를 지켜야 한다.

11. 여러개의 범주형 변수를 다룰때 묶은 막대를 사용하여 데이터를 시각화할 수 있다.
    묶은 막대 도표는 한 번에 많은 정보를 보여주는 만큼 헷갈리기도 쉽다. 따라서
    적절한 x,y축을 선택해야하고 하나에 담지 말고 기본 막대 그래프를 여러 개 그리는 것도 하나의 방법이다.

12. 인원 데이터를 남성/여성으로 이루어진 누적 막대 도표로 그릴 수 있다. 이 경우에는
    하나의 막대 그래프 안에 2개의 숫자만 들어가기 때문에 y축을 없애고 도표 안에
    숫자를 넣어서 더욱 간결하게 그릴 수도 있다.

13. 막대 도표로 요점을 전달하기 힘든 경우에는 점 도표, 히트맵을 사용하여 표현할 수 있다.
    각 나라의 인구(범주 2개)를 막대 그래프로 그리면 각 나라간 차이가 두드러지지 않아서
    보기가 힘든데 이때 점 도표를 사용할 수 있다.
    만약 각 나라의 연도별 인터넷 보급률(범주 3개)을 표현하려면
    히트맵(x : 기간, y : 국가, 색상 : 보급률)을 사용하는 것이 효과적이다.
```

```python
# 2. 데이터 분포의 시각화 : 히스토그램, 밀도 도표, 누적 분포 함수, QQ 도표
14. 데이터를 구간별로 나눠서 표시하는 시각화 방식을 히스토그램이라고 한다. 히스토그램의 모든 구간의 폭은 같으며
    구간의 폭을 데이터를 잘 표시할 수 있는 값으로 정하는 것이 중요하다. 최근에는 가우시안 커널을 활용한
    밀도 도표를 그리는 추세이며 밀도 도표는 히스토그램을 곡선으로 부드럽게 이은 버전이라고 생각하면 된다.
    밀도 도표의 대역폭 또한 히스토그램의 구간 폭처럼 여러 값으로 실험해보아야 하며 도표의 양쪽 끝단에
    데이터가 존재할 수 없는 영역까지 분포 곡선이 뻗어있지 않은지 확인해보아야 한다.

15. 둘 이상의 분포를 누적 히스토그램이나 중첩 히스토그램으로 표현할 수 있지만 이는 데이터를 파악하기 어렵게 한다.
    이때 중첩 밀도 도표를 사용하면 4~5개의 분포도 적절하게 시각화할 수 있다.

16. 경험적 누적 분포 함수(ECDF)와 QQ도표(Quantile-Quantile Plot)은 데이터의 분포를 시각화하기에 좋은 방법으로
    데이터가 방대한 경우에도 모든 데이터를 한 번에 시각화할 수 있다는 장점이 있다.
    ECDF는 X축이 점수 Y축이 %라고 했을때 상위 20%는 몇 점인지 등을 아주 쉽게 확인할 수 있게 해주고.
    QQ도표는 비대칭적인 분포를 가지는 데이터셋에서 데이터의 분포가 정규분포나 로그스케일을 정확히 따르는지
    등을 확인할 수 있다. (이론적 참조선에 데이터가 가깝게 놓여있을수록 이론에 부합한다는 뜻)

# 2-1 여러 분포 상태의 결합 시각화
17. 만약 12개월의 온도 분포를 시각화하려고 한다면, 이때는 박스 플롯, 바이올린 도표 등을 사용할 수 있다.
    이때 y축(기온)을 반응 변수(response variable)라고 하고, x축(월)을 그룹화 변수(grouping variable)이라고 한다.

18. 박스플롯에서 박스의 크기는 데이터의 50%를 설명하고 중앙의 선은 중간 값을 의미한다.
    박스 높이의 1.5배 만큼 위 아래로 뻗은 세로선을 수염(whisker)이라고 하며 선의 끝 부분을
    각각 upper fence, lower fence라고 하고 울타리를 벗어난 데이터를 이상치로 분류한다.

19. 현대에는 박스플롯 대신 바이올린 도표를 더욱 많이 사용하며, 박스플롯을 어떠한 경우에도 대체할 수 있고
    데이터의 미묘한 차이를 더 잘 보여주고, 쌍봉 데이터도 표현 가능하다는 장점이 있다.
    다만 밀도 추정 방식을 쓰는 만큼 데이터가 없는 영역이나 부족한 영역을 그렇지 않은 것 처럼 표현하는 경우도 있다.
    이런 단점을 보완하기 위해서는 데이터 포인트를 일일이 점 형태로 표현하는 스트립 차트(Strip chart)를
    사용할 수 있는데, 스트립 차트 또한 점이 많이 겹쳐 있으면 보기 어려워 지터링(jittering)을 사용해 점을 분산시킨다.
    마지막으로 지터링 된 스트립 차트와 바이올린 도표를 합친 도표를 시나 도표(sina plot)이라고 한다.

20. y축에 시간을 두고 융기선 도표를 사용해 시간의 흐름에 따른 분포의 변화 추세를 시각화할 수 있다.
    예를 들어 장기간 온도 분포의 변화나 시간에 따른 영화 상영시간 분포등을 표현하는데 사용한다.
```

```python
# 3. 비율 데이터의 시각화
21. 파이차트는 단순한 분수를 강조하거나 단일 데이터셋을 표현할때 시각화에서 이점을 가진다.(국회 정당 비율)
    그러나 여러개 데이터셋을 표현할때는 막대 그래프를 사용하는 것이 효과적인데 시간에 따른 남녀 비율처럼
    막대가 2개만 필요할때는 누적 막대 그래프가 효과적이고, 더 많다면 병렬 막대가 효과적이다.
    왜냐하면 여러개의 막대가 누적되어 있을때 각 막대의 높이를 비교하기가 매우 어렵기 때문이다.

22. 연령대별 건강상태(훌륭함,좋음,보통,나쁨)나 결혼 상태(미혼,기혼,이혼,사별) 비율 등을 표현하기 위해서는
    누적 밀도 도표보다는 y축을 전체 인원 x축을 연령으로 해서 여러 개의 밀도 도표를 그리는 것이 효율적이다.
    만약 몇 살쯤 기혼자가 50%를 넘어가는지 등이 궁금할때는 y축을 절대적인 숫자가 아닌 비율로 바꾸어서 표현하면 된다.

23. 만약 결혼 상태에 따른 건강 상태 비율이나 시대별 건축 재료 비율 등 여러 개의 범주형 변수를 활용하고자 한다면
    모자이크 도표나 트리맵을 사용할 수 있다. 모자이크 도표는 누적 막대 그래프와 유사하지만 막대 그래프의 폭이
    서로 다르다는 차이점이 있다. 트리맵은 X,y축 없이 전체 공간을 분할하여 각 범주의 비율을 시각화하며,
    여러 범주형 변수의 조합이 중요한 의미를 가지지 않을때 효과적이다.
    그러나 모자이크 도표, 트리맵은 서로 다른 직사각형의 크기를 만들어서 데이터의 크기 차이를 명확하게 비교하기
    힘들다는 한계가 있다. 따라서 가능하다면 실제 숫자나 백분율을 도표 위에 표시하는 것이 좋다.

24. 표현해야 하는 범주형 변수가 많아지면 모자이크 도표나 트리맵도 어수선해지기 마련이다.
    이때는 평행 집합 도표(parallel set plot)을 사용할 수 있는데, 각각의 분류 기준을 띠로 연결하고
    음영과 색을 부여해 구분한다. 이때 색을 칠하는 순서와 집합의 순서를 잘 배열해서 띠가 교차하는 지점이
    최소화되도록 하는 것이 중요하다.
```

```python
# 4. 여러 정량 변수의 관계 시각화(X-y)
25. 데이터 포인트의 두 변수 사이 관계(몸무게와 키)등을 파악하기 위해서 산점도(scatter plot)이 효과적이다.
    산점도는 포인트의 색상(성별)을 추가할 수 있으며 추가로 데이터 포인트의 크기(나이)등을 추가할 수도 있다.
    데이터 포인트의 크기를 추가한 것을 버블 차트라고 한다. 그러나 버블 차트는 변수의 상관관계를 적절하게
    나타내기 어려울 수 있는데, 이때는 산점도 매트릭스를 그려 관여하는 변수를 제한할 수 있다.

26. 변수 간 상관계수를 시각화한 것을 상관곡선(correrlogram)이라고 한다. 색상을 부여해 Matrix 형태로
    그릴 수 있으며, 추가적으로 상관관계의 강도를 원의 크기로 나타내서 훨씬 시각적으로 강조되도록 할 수 있다.

27. 만약 유사한 두 변수(x = 1970년 평균 수명, y = 2010년 평균 수명)을 비교하려면 가장 간단한 방법은
    x = y 사선 위에 산점도를 그리는 것이다. 전세계의 데이터를 모아 산점도를 그렸는데 x = y 대각선보다 위에 있는
    점이 많다면 실제로 평균 수명이 늘어났다고 시각적으로 쉽게 판단할 수 있기 때문이다.
    만약 데이터셋이 많지 않다면 경사 차트(slopegraph)가 더 유용하다. 경사 차트는 특히 여러개의 측정 값을
    비교할 수 있다는 장점이 있다.(1970~2010년의 데이터를 10년 단위로 이어서 그릴 수 있다.)

28. 시계열 데이터는 보통 선 그래프(점으로 표현된 데이터를 이웃하는 데이터끼리 선으로 이은 것)으로 표현한다.
    실제로는 점을 생략하기도 하고 데이터의 흐름을 돋보이게 하기 위해서 곡선 아래쪽에 색을 채우기도 한다.
    다만 이때는 y축의 시작점이 0일때만 사용 가능하다.

29. 반응 변수(response variable)가 2개 이상일때, 선 그래프를 2개 그릴수도 있지만 연결 산점도(connected scatterplot)를
    그려볼 수 있다. 공학에서는 위상도(phase portrait)라고도 불리는 이 방식은 두 변수의 상관관계를
    선 그래프로 표현한 것이라고 할 수 있다. 연결 산점도는 해석하기가 조금 어려운 측면이 있지만
    개별 선 그래프에서 파악할 수 없는 정보를 제공하기도 한다. 또한 연결 산점도를 그릴 때는 데이터의 방향과
    시간 범위를 모두 표시하는 것이 중요하다. (선의 색상을 점점 짙게 해 방향을 나타내고, 선의 일정 구간마다
    날짜를 표기해두어 정확하게 이해할 수 있도록 한다.)

# 4-1. 추세의 시각화
30. 시계열 데이터의 추세를 시각화하기 위해서는 보통 이동 평균을 사용하지만 길이를 길게 잡을수록 곡선이 짧아지며
    다소 울퉁불퉁하게 표현된다는 한계가 존재한다. 이를 해결하기 위해 일반화 가법 모형(GAM)을 사용할 수 있는데
    그 중 LOWESS(Locally Weighted Scatterplot Smoothing)는 데이터 포인트의 일정 부분에 대한 로컬 회귀 모델로
    계산 비용이 높지만 시계열 데이터 외에 임의의 산점도 모델에 모두 적용할 수 있다는 장점이 있다.
    그 외에도 스플라인(spline) 모델을 사용할 수 있는데 스플라인은 LOESS에 비해 계산 복잡도가 낮고
    다양한 변형을 가지고 있다는 장점이 있지만 데이터에 적합하지 않은 모델을 선택하면
    오히려 데이터 해석에 방해가 될 수 있다.

31. 종종 데이터의 x축 또는 y축(혹은 둘다)을 지수 변환하거나 로그 변환하는 것이 데이터 시각화에 도움이 될 때가 있는데
    이렇게 하면 곡선으로 표현되는 추세선을 직선으로 표현할 수 있다는 장점을 가지게 된다.
    y축이 로그, x축이 선형인 곡선을 로그-선형이라고 부르며 그 외에도 로그-로그, 선형-로그 등이 자주 쓰인다.

32. 시계열 데이터에 장기적으로 지배적인 추세가 있을때, 이를 제거하면 중요한 편차를 특정해서 강조하기 좋은데
    이를 추세 제거(detrending)이라고 한다. 예를 들면 우상향하는 주택 가격 그래프에서 우상향 추세를 제거하면
    주택 시장의 거품을 명확하게 파악할 수 있다.

33. 시계열 데이터는 크게 장기적 추세, 임의 노이즈, 외부 사건, 주기적인 변화로 나눌 수 있는데
    계절성 시계열 분해(STL)을 활용하면 이 네 가지를 분해해서 각각의 데이터의 영향을 파악할 수 있다.
```

```python
#5 불확실성의 시각화

34. 불확실성을 표현할 때는 가장 흔하게 오차 막대와 신뢰 대역을 사용하지만 직관적인 시각화를 위해서
    데이터의 정확도나 밀도를 조금 포기하고 빈도 프레이밍이나 애니메이션등을 사용할 수 있다.
    빈도 프레이밍(frequency framing)이란 확률을 빈도로 시각화하는 작업으로 무작위성을 시각적으로 강조할 수 있다.
    그러나 0과 1이 아닌 어떠한 확률 분포로 불확실성을 시각화해야 하는 경우도 있다. 확률분포 그래프 또는
    그 분포 곡선을 점으로 채운 분위수 점 도표를 그릴 수 있는데, 이것 모두 정확도를 희생하고 시각화를 강조한 것이다.

35. 다양한 한계에도 불구하고 불확실성을 시각화할때 오차막대를 많이 사용한다(특히 빈도주의에서)
    어떠한 데이터셋이 있을때 이 데이터의 평균, 표준편차, 표준오차, 신뢰구간 등을 오차 막대로 표현할 수 있는데
    표준 편차에 표본크기의 제곱근으로 나눈 근사값이 표준 오차이며 표준 오차의 구간을 2배로 늘리면 95% 신뢰 구간을 나타낸다.
    신뢰 구간은 80%, 95%, 99%를 묶어서 단계별 오차 막대로 표현할 수 있는데, 만약 어떤 데이터A 의 평균이
    다른 데이터B 보다 유의하게(Significant) 높다는 것을 증명하기 위해서는 A의 신뢰 구간 95% (유의 수준 5%)가
    데이터B의 평균보다 앞에 있어야 한다.

36. 빈도주의가 신뢰 구간을 통해 불확실성을 가늠하는 반면, 베이지안은 사후 분포와 신용 구간을 계산한다.
    이 둘의 가장 큰 차이점은 빈도 주의는 실제 모수 값이 0이 아니라고 얼마나 확신할 수 있을지를 말한다면
    베이지안은 실제 모수 값이 어디에 있다고 예상해야 할지를 찾는 것이라는 점이다.
    예를 들어 영화 A보다 영화 B의 평점이 높다는걸 증명하려고 할때,
    빈도주의에서는 영화 B의 평점에 대한 95% 신뢰 구간이(표준 오차 막대의 2배) 영화A의 평점 보다 완전히 높을때
    유의한 차이가 있다고 보는 것이고.
    베이지안에서는 영화A와 B의 평점이 같다는 사전 분포를 기반으로 영화 B의 평점 샘플의 사후 분포를 그렸을때
    그 분포도의 신용 구간 95%가 평점 A의 평균보다 높을때 유의한 차이가 있다고 보는 것이다.

37. 데이터셋의 추세의 불확실성을 표현하기 위해서는 신뢰 대역(confidence band)를 그릴 수 있다.
    선형 회귀(직선)이라고 할지라도 신뢰 대역은 곡선일 수 있으며 단계별 신뢰 대역을 그려 신뢰 수준을
    강조할 수 있다. 비선형 곡선도 마찬가지로 곡선 주위에 신뢰 대역이 형성되며 직선보다 굴곡이 심하다.

38. 불확실성을 시각화할 때는 항상 결정론적 해석 오류(deterministic construal error, 불확실성을 그 데이터의
    결정론적 특성이라고 해석하는 경향)이 나타날 수 있기 때문에 이 문제를 피하기 위해서는
    가설적 결과 도표(hypothetical outcome plot)이라는 시각화 방식을 사용할 수 있다. 이 방식은 가능도는 같지만
    서로 다른 도표들을 보여주는 애니메이션울 통해 불확실성을 표현하는 방법으로, 유용한 방법이지만
    각 도표에 들어가는 결과들이 실제 분포를 전형적으로 보여주도록 신경써서 설계해야 한다.
```
