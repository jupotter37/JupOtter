# **Data Visualization**

```python
import plotly.express as px
import plotly
```

```python
values1 = data[' Estimated Unemployment Rate (%)']
category = data['Region']
```

```python
fig = px.pie(data , values = values1 , names=category , title='Estimated Unemployment Rate (%)')
fig.update_traces(textposition="inside" , textinfo="percent")
fig.update_layout(autosize=False,
    width = 600,
    height = 600,
    title_font_size = 15
)
```

```python
values2=data[' Estimated Employed']
category=data['Region']
```

```python
fig = px.pie(data , values = values2 , names=category , title='Estimated Employed')
fig.update_traces(textposition="inside" , textinfo="percent")
fig.update_layout(autosize=False,
    width = 600,
    height = 600,
    title_font_size = 15
)
```

```python
values3 = data[' Estimated Labour Participation Rate (%)']
category = data['Region']
```

```python
fig = px.pie(data , values = values3 , names=category  , title='Estimated Labour Participation Rate (%)')
fig.update_layout(
    autosize = False,
    width = 600,
    height = 600
)
```
