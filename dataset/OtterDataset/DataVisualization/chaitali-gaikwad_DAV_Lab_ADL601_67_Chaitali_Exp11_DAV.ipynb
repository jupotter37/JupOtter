**Explore Top-5 Data Visualization Libraries in Python (w.r.t Features & Applications)**

1. Matplotlib: Matplotlib is a widely-used Python library for creating static, interactive, and publication-quality visualizations. It offers a vast array of plotting functions, allowing users to generate a variety of plots including line plots, scatter plots, histograms, and more. With its flexibility and customization options, Matplotlib is favored by scientists, engineers, and data analysts for data exploration and presentation.

2. Seaborn: Seaborn is a Python data visualization library that builds on top of Matplotlib and provides a high-level interface for creating attractive statistical graphics. It simplifies the process of generating complex visualizations such as distribution plots, categorical plots, and relational plots. Seaborn's ease of use and aesthetic appeal make it a valuable tool for data analysis and exploration.

3. Plotnine (ggplot): Plotnine is a Python implementation of the ggplot2 library in R, offering a grammar of graphics approach to visualization. It enables users to create complex plots by mapping data to visual elements in a concise and readable manner. Plotnine's adherence to the principles of the Grammar of Graphics facilitates reproducibility and clarity in code, making it a popular choice for creating publication-quality visualizations in Python.

4. Bokeh: Bokeh is a Python library for creating interactive visualizations that are designed to be displayed in modern web browsers. It provides a flexible and powerful framework for building interactive plots, dashboards, and data applications. Bokeh's interactive features, including widgets and streaming data capabilities, make it well-suited for developing dynamic and engaging web-based data visualizations.

5. Pygal: Pygal is a lightweight Python library for creating SVG-based charts and graphs. It offers a simple and intuitive API for generating various types of charts including line charts, bar charts, and pie charts. Pygal's focus on simplicity and scalability makes it a convenient choice for generating static and interactive visualizations that can be easily embedded into web applications or documents.

```r
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
from bokeh.transform import cumsum
import altair as alt
from sklearn.datasets import load_iris
from bokeh.palettes import Category20c
from math import pi
```

```r
# Load Iris dataset from scikit-learn
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
iris_target = iris.target
iris_target_names = iris.target_names
```

Perform simple experiments on the 2 identified libraries for Python

```r
# Scatter plot with Matplotlib
plt.figure(figsize=(10, 6))
for species, color in zip(df['species'].unique(), ['red', 'green', 'blue']):
    plt.scatter(df[df['species'] == species]['sepal length (cm)'], df[df['species'] == species]['sepal width (cm)'], c=color, label=species)
plt.title('Matplotlib Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```r
# Bar plot with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='species', y='sepal length (cm)', data=df, hue='species')
plt.title('Seaborn Bar Plot')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.legend(title='Species')
plt.show()
```

```r
#bokeh library
# Count occurrences of each species
species_count = pd.Series(iris_target_names[iris_target]).value_counts()

# Prepare data for pie chart
data = pd.DataFrame({'species': species_count.index, 'count': species_count.values})
data['angle'] = data['count']/data['count'].sum() * 2*pi
data['color'] = ['blue', 'green', 'red']

# Create the figure
p = figure(title="Iris Species Distribution", toolbar_location=None,
           tools="hover", tooltips="@species: @count", x_range=(-0.5, 1.0))

# Plot pie chart
p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='species', source=data)

# Customize plot attributes
p.axis.visible = False
p.grid.grid_line_color = None

# Show the plot
output_notebook()
show(p)


```

## **R libraries for data visualization**

**Explore Top-5 Data Visualization Libraries in R (w.r.t Features & Applications)**

1. RGL: RGL, or "R Graphics Library," is a powerful package for creating 3D visualizations in R. It provides functions to create interactive 3D plots and animations directly in R. With RGL, users can create complex 3D scenes, manipulate objects, apply lighting effects, and interactively explore data in three dimensions.

2. Plot3D: The plot3D library in R provides functions for creating 3D scatterplots, surfaces, and contour plots. It offers tools for visualizing multivariate data in a three-dimensional space, enabling users to explore relationships and patterns within their datasets.

3. Leaflet: Leaflet is an R package for creating interactive maps and web-based geographic visualizations. It integrates the Leaflet JavaScript library with R, allowing users to create dynamic and interactive maps directly in R scripts and Shiny applications. Leaflet provides functions to add layers, markers, pop-ups, and controls to maps, making it easy to customize the appearance and behavior of maps.

4. ggplot2: ggplot2 is a popular R package for creating elegant and versatile graphics based on the grammar of graphics. It provides a consistent and flexible syntax for building complex visualizations by mapping data to aesthetic attributes such as color, shape, size, and position. ggplot2 follows the layered approach to constructing plots, allowing users to add multiple layers of data and geometric objects to create sophisticated visualizations.

5. Lattice: Lattice is an R package for creating conditioned plots and trellis displays. It provides a powerful framework for visualizing multivariate data by creating panels or small multiples based on conditioning variables. Lattice plots are created using high-level functions that take formula syntax, making it easy to specify relationships between variables and create customized displays for different subsets of data.
