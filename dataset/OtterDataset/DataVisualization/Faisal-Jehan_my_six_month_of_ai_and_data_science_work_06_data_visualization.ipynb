# Data Viz

```python
# import libraries
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Import titanic data from seaborn
df=sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age',y='fare' )
```

```python
df.plot(kind='scatter', x='age',y='fare',color='red')
```

```python
df.plot(kind='scatter', x='age',y='fare',color='yellow')                                                                              #e8e809
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar' , color='yellow')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('AAverge Fare')
plt.xlabel('Gender')
```

```python
# seaborn plots
# bar plot using seaborn as df
sns.scatterplot( data=df , y='fare' , x='age', hue='class')
```
