```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x,y , c = colours , alpha=.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Question 1")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure with four subplots arranged in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# Plot the first line on the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# Plot the second line on the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# For line 3:
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# Plot the third line on the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# For line 4:
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plot the fourth line on the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Set the overall title for the figure
fig.suptitle('Four Line Plots')

# Display the plot
plt.show()
```
