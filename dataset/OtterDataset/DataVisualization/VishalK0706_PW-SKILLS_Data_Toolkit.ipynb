```python
# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset (e.g., Iris dataset)
df = sns.load_dataset('iris')

# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['number']).columns
corr_matrix = df[numerical_cols].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap of the correlation matrix using Seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, linewidths=0.5)

# Set the title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

```python
#7.Write a program a perform element-wise multiplication of Numpy arrays.

import numpy as np

# Create two NumPy arrays
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

# Method 1: Element-wise multiplication using the * operator
result1 = array1 * array2

# Method 2: Element-wise multiplication using numpy.multiply()
result2 = np.multiply(array1, array2)

# Display the results
print("Array 1:", array1)
print("Array 2:", array2)
print("Element-wise multiplication using * operator:", result1)
print("Element-wise multiplication using numpy.multiply():", result2)
```

```python
#8.Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y1 = np.sin(x)                # First line (sine wave)
y2 = np.cos(x)                # Second line (cosine wave)
y3 = np.tan(x)                # Third line (tangent wave)

# Create the line plot
plt.plot(x, y1, label='Sine', color='blue')      # Sine wave
plt.plot(x, y2, label='Cosine', color='orange')   # Cosine wave
plt.plot(x, y3, label='Tangent', color='green')    # Tangent wave

# Add titles and labels
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()

# Show the plot
plt.grid()  # Optional: Add grid for better readability
plt.ylim(-2, 2)  # Optional: Limit y-axis for tangent function visibility
plt.show()
```

```python
#10.Create a histogram using Seaborn to visualize a distribution.

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a histogram of body mass with KDE and hue by species
sns.histplot(data=penguins, x="body_mass_g", kde=True, hue="species", bins=30)

# Add title and labels
plt.title('Histogram of Body Mass of Penguins')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
#11.Perform matrix multiplication using NumPy.
# Import the NumPy library
import numpy as np

# Define two matrices
matrix1 = np.array([[1, 2],
                     [3, 4]])

matrix2 = np.array([[5, 6],
                     [7, 8]])

# Method 1: Using np.dot()
result_dot = np.dot(matrix1, matrix2)
print("Matrix multiplication using np.dot():")
print(result_dot)

# Method 2: Using np.matmul()
result_matmul = np.matmul(matrix1, matrix2)
print("\nMatrix multiplication using np.matmul():")
print(result_matmul)

# Method 3: Using the @ operator
result_operator = matrix1 @ matrix2
print("\nMatrix multiplication using @ operator:")
print(result_operator)

```
