## Correlation 

Let's see if there's any correlation between any variables the phenomenon of happiness in a particular country. (The closer the number is to 1 between variables, the more correlated the variables are.) Let's begin with 2017.

```python
hap17.corr()
```

```python
ml.figure(figsize=(20,10)) 
sn.heatmap(hap17.corr(), annot=True)
```

Judging off the happiness data for 2017, the highest correlated areas appear to be Health Life Expectancy with Economy GDP per Capita. This also includes Family and Economy GDP per Capita, followed by Family and Health Life Expectancy. A probably more interesting tidbit would be how Freedom and Trust/Government/Corruption seem to be correlated at 0.5.

### Now let's move on to 2016.

```python
hap16.corr()
```

```python
ml.figure(figsize=(20,10)) 
sn.heatmap(hap16.corr(), annot=True)
```

Like 2017, the numbers and points that correlate with each other are almost the same. However, it's interesting to note that correlated relationships like Family and Health are lower than they were compared to 2017. 

### Now let's move on to 2015.

```python
hap15.corr()
```

```python
ml.figure(figsize=(20,10)) 
sn.heatmap(hap15.corr(), annot=True)
```

Compared to 2017 and 2016, the same correlated relationships mentioned before have dropped here.  The correlation between Freedom and Trust has also dropped by 0.1. Not much of a change between the 3 years, however, it is interesting to note that Family, Health (Life Expectancy), Economy (GDP per Capita) have slightly increased over the years. On a side note, it will be interesting to see how these numbers come into play for the 2018 numbers given all the political news.
