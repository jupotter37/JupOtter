There are many other useful dataframe methods that provide more details about your dataframe that can be found in the documentation. Here are just a few:
- `.dtypes` prints datatypes of all columns in a DataFrame
- `values` returns all the values of the dataframe in an array format
- `.info()` provides a bird's eye view of column data types and missing values in a DataFrame
- `.describe()` returns a distribution of numeric columns in your DataFrame
- `.unique()` finds the number of unique values in a DataFrame column

Two more useful methods for data manipulating we will learn about are:
- `.apply()`: applies a function along an axis of a dataframe
- `.groupby()`: groups a dataframe based on a specified condition



```python
df.values
df['households'].unique()
```

```python
def compare(df):
  if df['population'] > df['total_bedrooms']:
    return 1
  elif df['population'] == df['total_bedrooms']:
    return 0
  else:
    return -1

df['bed_pop_compare'] = df.apply(compare, axis = 1)
df.head()
```

```python
grouped = df.groupby(by = ['bed_pop_compare'])
grouped.mean()
```

**EXCERCISE 2**
1. Create a new column in `df` that determines the average number of people per household on each block.
2. From your new column, categorize the house according to the following rules:
- if avg number of people per household between 0 and 2 (inclusive): *couple*
- if avg number of people per household between 3 and 5: *family*
- if avg number of people per household greater than 5: *tenants*
3. Give the mean of the median_house_value and median_house_income for each category

```python
#1
df['average per household'] = df['population']/df['households']
df.head()

#2
def categorize(df):
  if 0 <= df['average per household'] <= 2:
    return 'couple'
  elif 2 < df['average per household'] <= 5:
    return 'family'
  else:
    return 'tenants'

df['household category'] = df.apply(categorize, axis = 1)
df.head()

```

```python
#3
grouped2 = df.groupby(by = ['household category'])
grouped2['median_house_value', 'median_income' ].mean()
```

### Data Cleaning
Now that we have a general sense of what pandas is and how to use it, we can start exploring how to us it to clean data. Similar to the previous excercises, data cleaning in pandas is expedited with the use of methods.



```python
from numpy import NaN
snakes = pd.DataFrame([[NaN, 2, '4', 1, True], [4, 4, 2, NaN, True], [4,4,2, NaN, True], [7,4,6, 0, True], [8, 3, 4, 0, False]],
                      index = ['cobra', 'viper', 'viper', 'black adder', 'sidewinder'],
                      columns = ['health', 'damage', 'speed', 'shield', 'venom'])
snakes
```

Let's first get rid of all the duplicates

```python
snakes.drop_duplicates(inplace = True)
snakes
```

`isna()` can be used to determine if an element is a Null/missing value.

```python
snakes.isna()
```

There are several ways to deal with missing values:
- `fillna()`
- `dropna()`

```python
snakes_fill = snakes.fillna(value = 3)
snakes_fill
```

```python
snakes_drop = snakes.dropna(axis = 0) #axis = 0 drops the row with a NaN value. axis = 1 drops the column with a NaN value
snakes_drop
```

The following is a real dataset of airbnb listings and has the following columns:
- `listing_id`: The unique identifier for a listing

- `host_id`: Unique identifier for a host
- `host_name`: Name of host
- `neighbourhood_full`: Name of boroughs and neighbourhoods
- `coordinates`: Coordinates of listing _(latitude, longitude)_
- `Listing added`: Date of added listing
- `room_type`: Type of room
- `rating`: Rating from 0 to 5.
- `price`: Price per night for listing
- `number_of_reviews`: Amount of reviews received
- `last_review`: Date of last review
- `reviews_per_month`: Number of reviews per month
- `availability_365`: Number of days available per year
- `Number of stays`: Total number of stays thus far

Go through this dataset, and try cleaning it for yourself.


```python
airbnb = pd.read_csv('https://github.com/adelnehme/python-for-spreadsheet-users-webinar/blob/master/datasets/airbnb.csv?raw=true', index_col = 'Unnamed: 0')
airbnb
```

```python
#fix missing values
airbnb.dropna(axis = 0, inplace = True)
```

```python
#fix that the description of the airbnb is under 'name'
airbnb.rename(columns = {'name':'description'}, inplace = True)
```

```python
#5_stars column more accurate
airbnb.rename(columns = {'5_stars':'5_stars(%)'}, inplace = True)
airbnb['5_stars(%)'] = airbnb['5_stars(%)'] * 100
```

```python

airbnb['room_type'].values
```

```python
airbnb['room_type'].unique()

```

```python
#room_type descriptions are not constant, now there will only be
airbnb['room_type'].unique()
airbnb['room_type'] = airbnb['room_type'].str.lower()
airbnb['room_type'] = airbnb['room_type'].str.strip()
room_mappings = {'private room': 'Private room',
                 'private': 'Private room',
                 'entire home/apt': 'Entire Home',
                 'home': 'Entire Home',
                 'shared room': 'Shared Room'}

airbnb['room_types'] = airbnb['room_type'].replace(room_mappings)
airbnb['room_types'].unique()
```

```python
#one row has a rating greater than 5
airbnb['rating'] = airbnb['rating'].where(airbnb['rating'] < 5, 5) #where condition is true, it will keep original value. where it is false it will replace
airbnb.describe()
```

```python
#split into neighbourhood and boroughs separately

neighbour_split = airbnb['neighbourhood_full'].str.split(',', expand = True)
airbnb['neighbour'] = neighbour_split[0]
airbnb['borough'] = neighbour_split[1]
airbnb.drop(['neighbourhood_full'], axis = 1, inplace = True)
```

```python
#longitude and latitude split
airbnb['coordinates'] = airbnb['coordinates'].str.replace('(','')
airbnb['coordinates'] = airbnb['coordinates'].str.replace(')','')
airbnb['coordinates'] = airbnb['coordinates'].str.replace(',','')
coordinates_split = airbnb['coordinates'].str.split(expand = True)
latitude = coordinates_split[0]
longitude = coordinates_split[1]
airbnb['latitude'] = latitude.astype(float)
airbnb['longitude'] = longitude.astype(float)
airbnb.drop(['coordinates'], axis = 1, inplace = True)
```

```python
#price has a dollar sign and is a string
airbnb['price'] = airbnb['price'].str.replace('$','').astype(float)
```

```python
#last review and listing added don't match up
airbnb['listing_added'] = pd.to_datetime(airbnb['listing_added'], format = '%Y-%m-%d')
airbnb['last_review'] = pd.to_datetime(airbnb['listing_added'], format = '%Y-%m-%d')
airbnb = airbnb[airbnb['listing_added'] <= airbnb['last_review']]
```
