## Basics on data visualization and plotting

### Why we need good data visualization?
A good figure is worth more than one thousand words. It is even more true in engineering and science.  As an engineer or scientist, you often need to show your design, a concept, calculation results, or measurement data to convey a message or transfer information. A good, professional visual or plot can help get better grade in class or stand out in competitions. "Good" data visualization also shows your deep understanding of the subject, creative thinking, rigor, and quality of the work. 

### What is "good" data visualization?
The standard for "good" is always subjective. However, there are some common traits in convincing and engaging visualizations and plots. There are many researches on human perception and I am sure you can easily find great literature on this topic. The goal here is to synthesize on some best practices for creating good data visuals.

* Always keep in mind the question "what is the point?" The main message should be front and center when people see your plot. For example,
    * show the relationship between time and the distance traveled by a car
    * show global warming by plotting the temperature record
    * show the correlation between two variables 
* Clear and simple. Try to make just one point in one figure. Don't pile things up.
* Basic elements of a plot:
    * axis, ticks, axis labels, legend, title, grid, unit, etc.
    * not all elements are necessary
    * can add text for annotation. But don't over do it. 

Show the [anatomy of a figure](https://matplotlib.org/gallery/showcase/anatomy.html).
   
* Choose the right font type and size:
    * Use simple and consistent font types: Try not to mix them. Most publishers specify font types such as New Tims Roman.
    * Use proper font size: not too large nor too small. 
* Choose the right plot type:
    * line chart, scatter plot, bar chart, pie chart, etc.
    * Color map
    * Contour map
    * quiver plot
    * 2D or 3D?
* Choose the right color and shade:
    * Color has meanings: for example, if you plot a temperature contour map, better use red for warm and blue for cool. It will confuse people if you make it opposite. 
    * Can use shade too
    * Color contrast and blending
    * Show the color map if you use color for quantitative data
    * Do you really need color? Can black/white or gray scale do the same job? Does your publisher/printer allow color? Remember color figures printed in black/white may loose the main message.
* Be creative and sometime it needs some imagination.

This section is general. It applies to data visualization and plotting using any tool or software, for example MS Excel, Matlab, or Python. The main differences between MS Excel plotting and Python plotting:

## Two modes of using Matplotlib

As mentioned earlier, Matplotlib was created to mimic Matlab, mainly through its PyPlot interface. This is only one mode of use Matplotlib and it is simple. The second mode is to its object-oriented interface which is used for more complex drawing. In general, it is recommended to use the object-oriented interface.

* pyplot interface. So far, all the examples shown use the pyplot interface and they indeed look like Matlab. With the pyplot interface, it is stateful, which means the code tracks the current figure (and other objects) which is created by the most recent "plt.plot()" function. You can create multiple figures and call "plot()" several times in sequence. The problem with this pyplot interface is that it is hard to update previous (not current) figures.

* Object-oriented interface. In this mode, you can create and record figures (and other things) as objects (variables). And then later on, you can call functions of these objects to make changes. The following example of creating two subfigures shows the idea.

But before executing the following cell, now it is probably the best place to introduce some key concepts which make the hierarchy of the Matplotlib plot: figure and axes.
* Figure: it is the outermost container for a matplotlib graphic, which may include several Axes objects. 
* Axes: it is an individual plot or graph. Don't confuse "Axes" with "axis". 
So the hierarchy looks like this. The Figure object is like a box which contains one or more Axes (plots). Below each Axes, there are other lower level objects such as tick marks, legends, text boxes, label, etc. The following figure shows this basic concept of object hierarchy in Matplotlib:

<img src="figure_hierarchy.png" alt="hierarchy" width="400"/>
([figure source:](https://realpython.com/python-matplotlib-guide/))



````python
# create a figure with two subplots
# fig will be the big box container for everything
# ax will be an array of two Axes objects, each of which corresponding to one subplot
fig, ax = plt.subplots(2)

# prepare some data for plot
x = np.linspace(0, 6.28, 30)
y1 = np.sin(x)
y2 = np.cos(x)

# Call plot() function on each subplot (Axes object)
ax[0].plot(x, y1)
ax[1].plot(x, y2)

# Now you can easily update/manipulate the subplot in whatever order you like
ax[0].set_ylabel("sin(x)")
ax[1].set_ylabel("cos(x)")
ax[1].set_xlabel("x")

plt.plot()
````

## Read and plot data from file

In engineering, you often need to plot some given data, such as data from measurements. These data may be stored in a file in simple text, CSV, or other format. There are many different ways to load data into python and plot them. Here we introduce two methods:
* use the built-in csv module to load CSV files
* use the Numpy's loadtxt function to load text files

The following example use the built-in csv module's CSV reader. CSV stands for "comma-separated values". CSV is a simple file format used for tabular data, such as a spreadsheet. Data files in the CSV format can be imported to and exported from programs such as Microsoft Excel. The rows of tabular data in CSV files are often delimited by a comma or space. It is noted that the CVF module does not require the data file have a "csv" as its file name extension. It only requires the data is delimited. 

This example first loads in the CSV file named "experimental_data.csv" which has two columns. The CSV reader can read one line at at time and then get the column data separated by the delimiter. Then, we can append each of the column data to the arrays "x" and "y", which can be used for plotting.

````python
import matplotlib.pyplot as plt
import csv

#declare empty arrays used to store data
x = []
y = []

with open('experimental_data.csv','r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # This skips the first (header) row of the CSV file.
    # If your CSV data file does not have header, then comment out this line
    next(csvreader)
    
    for row in csvreader:
        x.append(float(row[0]))
        y.append(float(row[1]))

plt.scatter(x,y, marker = 'o', label='Experimental data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of data read from CSV file')
plt.legend()
plt.show()
````

The second method is to use the Numpy's loadtxt() function. It does not require the data file to have the extension "txt". In fact, you can read in the same "csv" file we have for the previous example. 

````python
import matplotlib.pyplot as plt
import numpy as np

#directly load the data from file and assign the columns to "x" and "y"
#notice the "skiprows=1" because the first row is the header.
x, y = np.loadtxt('experimental_data.csv', delimiter=',', unpack=True, skiprows=1)

#the same plot function as above (with the exception of color)
plt.scatter(x,y, color='red', marker = 'o', label='Experimental data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of data read with Numpy loadtxt function')
plt.legend()
plt.show()
````

A more complete example which customizes the following:
* font types
* font size
* axis limit
* ticks
* legend location, font size, and frame
* ...

````python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)  #allow the use of Latex for math expressions and equations
plt.rc('font', family='serif') #specify the default font family to be "serif"

#directly load the data from file and assign the columns to "x" and "y"
x, y = np.loadtxt('experimental_data.csv', delimiter=',', unpack=True, skiprows=1)

#fit a linear curve
z = np.polyfit(x, y, 1) #Numpy's polynomial fit (with degree=1 which is a linear fit)
p = np.poly1d(z)        #convert it to one-dimensional polynomial (e.g., 3x^2+2x+2)

#scatter plot
plt.scatter(x, y, color='blue', marker = 'o', label='Experimental data')

#line plot for the fitted linear curve
plt.plot(x, p(x), color='red', label='Linear regression')

#set the limit for the x and y axes
plt.xlim([0,1.0])
plt.ylim([5,45])

#set x and y axes label and font size
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)

#show the ticks on both axes and set the font size 
plt.tick_params(axis='both', which='major', labelsize=12)

#show title and set font size
plt.title('Plot of data read with Numpy loadtxt function',fontsize=16)

#show legend, set its location, font size, and turn off the frame
plt.legend(loc='lower left',fontsize=14,frameon=False)
plt.show()
````

### Plot of vector data

In addition to line, scatter, and contour plots, you can also use Python to plot vector data, such as velocity, acceleration, and displacement. Although Python is very powerful in making these vector plots, it is by no means the best. There are other tools and software for plotting more complicated vector data, such as Paraview. 

Here, we show a simple example. Assume that we know a two-dimensional velocity field, which follows
\begin{equation}
U = x^2
\end{equation}
\begin{equation}
V = -2xy
\end{equation}
where $U$ and $V$ are velocity in the $x$ and $y$ directions, respectively. To visualize how the velocity field looks like, we can use the "quiver" function. The following code shows that the flow on the left half diverges when it hits the center plane and the flow on the right half converges and goes away from the center plane. 

````python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

plt.rc('text', usetex=True)  #allow the use of Latex for math expressions and equations
plt.rc('font', family='serif') #specify the default font family to be "serif"

x, y = np.meshgrid(np.arange(-10, 10, 1.5), np.arange(-10, 10, 1.5))
U = x**2
V = -2*x*y

plt.figure()
plt.title('Velocity field')
Q = plt.quiver(x, y, U, V, units='width')
#qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
#                   coordinates='figure')

#set the limit for the x and y axes
plt.xlim([-10,10])
plt.ylim([-10,10])

#set x and y axes label and font size
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)

#show legend, set its location, font size, and turn off the frame
#plt.legend(loc='lower left',fontsize=14,frameon=False)
#plt.savefig("velocity_acceleration_example.png",dpi=300, bbox_inches='tight', pad_inches=0)

plt.show()
````

With the analyical expressions for the velocity field, it is easy to calculate the acceleration field (hint: material derivative), which can be written as
\begin{equation}
a_x = 2x^3
\end{equation}
\begin{equation}
a_y = 2x^2y
\end{equation}

````python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

plt.rc('text', usetex=True)  #allow the use of Latex for math expressions and equations
plt.rc('font', family='serif') #specify the default font family to be "serif"

x, y = np.meshgrid(np.arange(-10, 10, 1.5), np.arange(-10, 10, 1.5))
a_x = 2*x**3
a_y = 2*x**2*y

plt.figure()
plt.title('Acceleration field')
Q = plt.quiver(x, y, a_x, a_y, units='width')
#qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
#                   coordinates='figure')

#set the limit for the x and y axes
plt.xlim([-10,10])
plt.ylim([-10,10])

#set x and y axes label and font size
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)

#show legend, set its location, font size, and turn off the frame
#plt.legend(loc='lower left',fontsize=14,frameon=False)
#plt.savefig("velocity_acceleration_example_2.png",dpi=300, bbox_inches='tight', pad_inches=0)

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
from skimage import measure
from mpl_toolkits.mplot3d import Axes3D

nx, ny, nz = (100, 100, 100)
x = np.linspace(-1.4, 1.4, nx)
y = np.linspace(-1.4, 1.4, ny)
z = np.linspace(-1.4, 1.4, nz)
xv, yv, zv = np.meshgrid(x, y, z)

F = (xv**2+9.0/4.0*yv**2.0+zv**2.0-1.0)**3.0 - (xv**2.0)*(zv**3.0)-9./80.0*yv**2*zv**3

verts, faces, _, _ = measure.marching_cubes_lewiner(F, 0, spacing=(0.1, 0.1, 0.1))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(verts[:, 0], verts[:,1], faces, verts[:, 2], color='red', linewidth=0, antialiased=False)
ax.view_init(30, 0)
plt.axis('off')
plt.show()
````
