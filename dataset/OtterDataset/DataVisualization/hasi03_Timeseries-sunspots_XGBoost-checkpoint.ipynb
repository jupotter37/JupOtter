# Model

```python
X, y =  group2.loc[:,['Month', 'Year']].values, group2.loc[:,'Number of Sunspots'].values
data_dmatrix = xgb.DMatrix(X,label=y)


```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
reg_mod = xgb.XGBRegressor(objective='reg:squarederror',
    n_estimators=1000,
    learning_rate=0.10,
    subsample=0.5,
    colsample_bytree=1, 
    max_depth=10,
)
reg_mod.fit(X_train, y_train)
```

```python
predictions = reg_mod.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print("RMSE: %f" % (rmse))
```

```python
# Creating an empty Dataframe with column names only
result = pd.DataFrame(columns=['Time', 'Test', 'Predicted'])
result['Time'] = pd.date_range(start='6/1/%i'%int(X_test[0,1]), periods=y_test.size, freq='M')
result['Test'] = y_test
result['Predicted'] = predictions


#Using Plotly to build the graph
fig = go.Figure()
fig.add_trace(go.Scatter(x=result['Time'][0:50], y=result['Test'][0:50],
                    mode='lines',
                    name='Test'))
fig.add_trace(go.Scatter(x=result['Time'][0:50], y=result['Predicted'][0:50],
                    mode='lines',
                    name='Predicted'))

# Edit the layout
fig.update_layout(title='Test vs Predicted Values',
                   xaxis_title='Months',
                   yaxis_title='Mean Sunspot Values',
                   template='gridon')
fig.show()
```
