# Data Visualization 

# 1.Bar Plot - Brand distribution 

```python
sns.countplot(x='Brand',data=mel)
plt.title("Brand distribution")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

This reveals that the brand distribution in the data is almost consistent 

# 2.Pie Plot - Manufacturing Year 

```python
manufacturing_years = mel['Year'].value_counts()
plt.pie(manufacturing_years,labels= manufacturing_years.index,autopct="%1.1f%%")
plt.title("Manufacturing years")
plt.show()
```

This shows that majority of the vehicles manufactured , were produced in 2017 and 2019
