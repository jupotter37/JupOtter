# **DATA VISUALIZATION**

```python
# Replacing the brackets from the year column and handling NaN values
df_movie['Year'] = df_movie['Year'].astype(str).str.replace(r'[()]', '', regex=True)
df_movie['Year'] = pd.to_numeric(df_movie['Year'], errors='coerce').fillna(0).astype(int)
```

```python
#Remove the min word from the 'Duration' and convert all values to numeric
df_movie['Duration'] =df_movie['Duration'].astype(str).str.replace('min','')
df_movie['Duration'] = pd.to_numeric(df_movie['Duration'], errors='coerce').fillna(0).astype(int)
```

```python
#splitting the Genre by , to keep only unique generes and replacing the null values with mode
df_movie['Genre']= df_movie['Genre'].str.split(',')
df_movie = df_movie.explode('Genre')
df_movie['Genre'].fillna(df_movie['Genre'].mode()[0] , inplace =True)
```

```python
#converts 'votes' to numeric and replace the , to keep only numerical part
df_movie['Votes'] = df_movie['Votes'].astype(str).str.replace(',' , '')
df_movie['Votes'] = pd.to_numeric(df_movie['Votes'], errors='coerce').fillna(0).astype(int)
```

```python
#Checking the dataset is there any null values present and data types of the features present
df_movie.info()
```

```python
 #Here we have created a histrogram over the years in the data
 year = px.histogram (df_movie,x="Year", histnorm= 'probability density', nbins=30)
 year.show()
```

```python
#Group data by Year and Calculate the average rating
avg_rating_by_year = df_movie.groupby(['Year' , 'Genre'])['Rating'].mean().reset_index()

#get top 10 genres
top_genres= df_movie['Genre'].value_counts().head(10).index

#Filter the data to include only th etop 3 Genres
average_rating_by_year= avg_rating_by_year[avg_rating_by_year['Genre'].isin(top_genres)]

#Create the line plot with plotly express
fig = px.line(average_rating_by_year, x='Year' , y='Rating' , color = 'Genre', title= 'Average' )

# Updating the detils into chart like title and hue
fig.update_layout(title = " Average Rating by Year for the top Genres" , xaxis_title= 'Year' , yaxis_title = 'Average Rating')

#Show the Chart
fig.show()
```

# **MODEL BUILDING**

```python
#Importing essential libraries for model building
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
model = LinearRegression()
from sklearn.metrics import accuracy_score , mean_absolute_error, mean_squared_error,r2_score
```

```python
#Grouping the columns with their average rating and then creating a new feature

genre_mean_rating = df_movie.groupby('Genre')['Rating'].transform('mean')
df_movie['Genre_mean_rating']= genre_mean_rating

director_mean_rating =df_movie.groupby('Director')['Rating'].transform('mean')
df_movie['Director_encoded']= director_mean_rating

actor1_mean_rating = df_movie.groupby('Actor 1')['Rating'].transform('mean')
df_movie['Actor 1_encoded'] = actor1_mean_rating

actor2_mean_rating = df_movie.groupby('Actor 2')['Rating'].transform('mean')
df_movie['Actor 2_encoded'] = actor2_mean_rating

actor3_mean_rating = df_movie.groupby('Actor 3')['Rating'].transform('mean')
df_movie['Actor 3_encoded'] = actor3_mean_rating
```

```python
#Keeping the predictor and target variable

x= df_movie[[ 'Year', 'Votes' , 'Duration' , 'Genre_mean_rating' , 'Director_encoded', 'Actor 1_encoded' , 'Actor 2_encoded' , 'Actor 3_encoded']]
y=df_movie['Rating']

#splitting the dataset into training and testing parts
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Building machine learning model and training them
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(x_train,y_train)
model_pred = model.predict(x_test)
```

```python
#Evaluating the performance of model with evaluation metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
print('The performance evaluation of Logistic Regression is below: ', '\n')

print ('Mean squared error: ' , mean_squared_error(y_test, model_pred))

print('Mean absolute error: ', mean_absolute_error(y_test, model_pred))

print('R2 score: ',r2_score(y_test, model_pred))
```
