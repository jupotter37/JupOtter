The following are the methods involved in retreiving specific data.

* df[]: To Select columns.
* df.loc[]: To Access a group of rows and columns by labels.
* df.iloc[]: To Access a group of rows and columns by integer positions.
* df.at[]: To Access a single value for a row/column label pair.
* df.iat[]: To Access a single value for a row/column pair by integer positions.

We will now explore each of these examples.

```python
#The following code displays a single column.
diabetes['BloodPressure'].head()
```

```python
#The following code displays multiple columns.
diabetes[['Pregnancies','BloodPressure','Glucose']].head()
```

```python
#To access a single row using loc function
diabetes.loc[2]
```

```python
#To access multiple rows using loc function
diabetes.loc[1:5]
```

```python
#To access multiple rows with specific columns using loc function
diabetes.loc[1:5, ['BMI', 'Age']]
```

```python
#To access a single row using iloc function
diabetes.iloc[3]
```

```python
#To access multiple rows using iloc function
diabetes.iloc[1:5]
```

```python
#To access multiple rows with specific columns using loc function
diabetes.iloc[1:10, 1:5]
```

We can understand a minute difference between loc and iloc by the example above. While displaying the rows 1:5, the loc function displayed records from index 1 to 5(inclusive of 5) whereas the iloc function displayed records from index 1 to 4(exclusive of 5).

```python
#The following code displays value at row index 1, column 'BloodPressure'
diabetes.at[1, 'BloodPressure']
```

```python
#The following code displays value at row position 1, column position 2(BloodPressure)
diabetes.iat[1, 2]
```
