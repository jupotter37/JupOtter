### Importing Libraries

```python
# Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

### Importing Dataset

```python
# Importing the required dataset
df = pd.read_csv('Dataset .csv')
```

```python
# getting first five rows of the dataset
df.head()
```

##### **Checking Number of Rows and Columns**

```python
# Get number of rows and columns
rows, cols = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
```

##### **Check for Missing Values in Each Column and Handle Them Accordingly**

```python
# Check for missing values in each column

missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# For categorical columns, you can either fill missing values with the mode (most frequent value) or drop them

df['Cuisines'].fillna(df['Cuisines'].mode()[0], inplace=True)
```

```python
# Verify missing values after handling them

print("Missing values after handling:")
print(df.isnull().sum())
```

#### **Visualize the Distribution of Ratings**

We will create a histogram to visualize the distribution of the Aggregate rating.

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Plot the histogram for the 'Aggregate rating'
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'].dropna(), bins=20, kde=True, color="blue")

# Add title and labels
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

**Note -** Most of rating is given between **3 to 4**.

#### **Compare the Average Ratings of Different Cuisines or Cities**

##### **Compare Top 10 Average Ratings of Different Cuisines**

```python
# Group by 'Cuisines' and calculate the mean of 'Aggregate rating'
cuisine_avg_rating = df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
```

```python
# Sort by average rating in descending order
cuisine_avg_rating_sorted = cuisine_avg_rating.sort_values(by='Aggregate rating', ascending=False).head(10)
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create a bar plot for the average rating of each cuisine
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Cuisines', data=cuisine_avg_rating_sorted, palette="coolwarm")

# Add title and labels
plt.title('Average Rating by Cuisine')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')

# Show the plot
plt.show()
```

**Note -** Top 10 Cuisine having average rating is **4.9**.

##### **Compare Top 10 Average Ratings of Different Cities**

```python
# Group by 'City' and calculate the mean of 'Aggregate rating'
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().reset_index()
```

```python
# Sort by average rating in descending order
city_avg_rating_sorted = city_avg_rating.sort_values(by='Aggregate rating', ascending=False).head(10)
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create a bar plot for the average rating of each city
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='City', data=city_avg_rating_sorted, palette="coolwarm")

# Add title and labels
plt.title('Average Rating by City')
plt.xlabel('Average Rating')
plt.ylabel('City')

# Show the plot
plt.show()
```

**Note -** From above graph we can see that, **Inner city** has maximum average rating

#### **Visualize the Relationship Between Various Features and the Target Variable**

##### **1. Relationship between Price Range and Aggregate Rating**

```python
# Plot a boxplot to see how 'Price range' affects 'Aggregate rating'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette="coolwarm")

# Add title and labels
plt.title('Relationship between Price Range and Aggregate Rating')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')

# Show the plot
plt.show()
```

**Note -** From above graph we can see that, The **1st** Price range having maximum aggregate rating.

##### **2. Relationship between Number of Votes and Aggregate Rating**

```python
# Plot a scatter plot to see how 'Votes' affect 'Aggregate rating'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, color='blue')

# Add title and labels
plt.title('Relationship between Votes and Aggregate Rating')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')

# Show the plot
plt.show()
```

**Note -** From above graph we can see that, Maximum Aggregate Ratings gives zero votes.

##### **3. Relationship between Has Table Booking and Aggregate Rating**

```python
# Convert 'Has Table booking' to a categorical variable (0 or 1)
df['Has Table booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})

# Plot a boxplot to see how 'Has Table booking' affects 'Aggregate rating'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=df, palette="coolwarm")

# Add title and labels
plt.title('Relationship between Table Booking and Aggregate Rating')
plt.xlabel('Has Table Booking')
plt.ylabel('Aggregate Rating')

# Show the plot
plt.show()
```

**Note -** From above graph we can see that, The customers who not table booked having maximum aggregate ratings.
