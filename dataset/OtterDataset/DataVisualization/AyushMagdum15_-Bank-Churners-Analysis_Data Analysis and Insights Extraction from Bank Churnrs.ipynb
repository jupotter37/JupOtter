Visualizations

```python
sns.histplot(x='Customer_Age',data=df)
plt.title("Age wise Frequency")
```

```python
#attrition wise grouping of customers
attrition_wise=df.groupby(['Customer_Age','Attrition_Flag']) ['CLIENTNUM'].count().reset_index()
attrition_wise
```

```python
plt.figure(figsize=(18,8))
sns.barplot(x='Customer_Age',y='CLIENTNUM',hue='Attrition_Flag',data=attrition_wise)
```

```python
#attrition wise and gender wise grouping of customers
gender_wise=df.groupby(['Attrition_Flag','Gender']) ['CLIENTNUM'].count().reset_index()
gender_wise
```

```python
sns.barplot(x='Gender',y='CLIENTNUM', hue='Attrition_Flag',data=gender_wise)
plt.legend(bbox_to_anchor=(2,1),loc='upper right')
```

```python
df2=df.groupby(['Attrition_Flag','Education_Level']) ['CLIENTNUM'].count().reset_index()
df2
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot(x='Attrition_Flag',y='CLIENTNUM',hue='Attrition_Flag',style='Education_Level',data=df2)
```

```python
#checking for outliers
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition_Flag', y='Total_Trans_Amt', data=df)
plt.title('Outliers in Total Transaction Amount')
plt.show()
```

```python
age_mean=df['Customer_Age'].mean()
age_median=df['Customer_Age'].median()
age_mode=df['Customer_Age'].mode()[0]
age_std=df['Customer_Age'].std()
print(f'Age-Mean:{age_mean},Median:{age_median},Mode:{age_mode},Std Dev:{age_std}')
```
