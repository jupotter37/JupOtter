```python
def analizar_dataframe(df):
    # Obtener información básica sobre el dataframe
    print("Información básica del dataframe:")
    print(df.info())
    
    # Obtener estadísticas descriptivas de las variables numéricas
    print("\nEstadísticas descriptivas de las variables numéricas:")
    print(df.describe())
    
    # Obtener los valores únicos de las variables categóricas
    print("\nValores únicos de las variables categóricas:")
    for columna in df.select_dtypes(include=["object"]).columns:
        print(f"{columna}: {df[columna].unique()}")
    
    # Verificar si hay valores nulos en el dataframe
    if df.isnull().sum().sum() > 0:
        print("\nValores nulos en el dataframe:")
        print(df.isnull().sum())
    else:
        print("\nNo hay valores nulos en el dataframe.")
```

```python
df = coursera
```

```python
print(analizar_dataframe(df))
```
