```python
import re
def extract_start_year(x):
    if x == "" or pd.isnull(x):
        return None  # or any other value you prefer for empty strings
    else:
        match = re.search(r'\d+', x)
        if match:
            return int(match.group())
        else:
            return None
        
pipeline = mdf['Pipeline'].apply(lambda x: extract_start_year(x))
start = mdf['When did you start this company?'].apply(lambda x: extract_start_year(x))

mdf['Age at Application'] = pipeline - start
mdf = mdf[mdf['Age at Application'] >= 0]
```
