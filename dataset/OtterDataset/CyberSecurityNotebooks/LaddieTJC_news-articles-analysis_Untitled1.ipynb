```python
model = AutoModelForSeq2SeqLM.from_pretrained("hyunwoongko/ctrlsum-cnndm")

tokenizer = PreTrainedTokenizerFast.from_pretrained("hyunwoongko/ctrlsum-cnndm")
def ctrl_summarize(text,prompt=None):
    if prompt:
        input_ids = tokenizer(prompt+'=>'+text, return_tensors="pt").input_ids
    else:

        input_ids = tokenizer(text, return_tensors="pt").input_ids
    output_ids = model.generate(input_ids,max_new_tokens=512)
    # Get the summary from the output tokens
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)
```
