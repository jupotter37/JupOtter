```python
pdf_folder_path = '/content/gdrive/My Drive/hack'
docs = []
for file in os.listdir(pdf_folder_path):
    if file.endswith('.pdf'):
        pdf_path = os.path.join(pdf_folder_path, file)
        loader = PDFPlumberLoader(pdf_path)
        docs.extend(loader.load())
```

```python
# Function to extract text from PDF files using pdfplumber
def extract_text_from_pdfs(pdf_folder_path):
    raw_text = ""
    pdf_docs = [os.path.join(pdf_folder_path, fn) for fn in os.listdir(pdf_folder_path) if fn.endswith('.pdf')]
    for pdf in pdf_docs:
        with pdfplumber.open(pdf) as pdf_file:
            for page in pdf_file.pages:
                raw_text += page.extract_text()
    return raw_text


pdf_folder_path = '/content/gdrive/My Drive/hack/'


raw_text = extract_text_from_pdfs(pdf_folder_path)


print(raw_text)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(docs)
vectorstore = Chroma.from_documents(documents=splits, embedding=OpenAIEmbeddings())

# Retrieve and generate using the relevant snippets of the blog.
retriever = vectorstore.as_retriever()
prompt = hub.pull("rlm/rag-prompt")
llm = ChatOpenAI(model_name="gpt-4", temperature=0)


def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)


rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
# Streamlit App
st.title("Onboarding Employee Helper")
st.write("Ask any question regarding employee onboarding.")

question = st.text_input("Your Question:")

if question:
    with st.spinner("Fetching answer..."):
        answer = rag_chain.invoke(question)
        st.write("Answer:", answer)
```

```python
from langchain_core.runnables import RunnableParallel

rag_chain_from_docs = (
    RunnablePassthrough.assign(context=(lambda x: format_docs(x["context"])))
    | prompt
    | llm
    | StrOutputParser()
)

rag_chain_with_source = RunnableParallel(
    {"context": retriever, "question": RunnablePassthrough()}
).assign(answer=rag_chain_from_docs)

rag_chain_with_source.invoke("Can I drink while working?")
```
