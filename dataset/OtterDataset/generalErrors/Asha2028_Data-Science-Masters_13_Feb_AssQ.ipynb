Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans:

The ArithmeticError Exception is the base class for all errors associated with arithmetic operation.
    Like:
    
    ZeroDivisionError
    
    OverflowError
    
    FloatingPointError

```python
# Example - OverflowError
a = 100.0

try: 
    for i in range(1,1000):
        a = a**i
        print(a)
except ArithmeticError as e:
    print("{}".format(e.__class__))
```

```python
#Example - ZeroDivisionError

try:
    1/0
except ArithmeticError as e:
    print(e,e.__class__)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans:

Exception LookupError is the base class for those exceptions that are raised when a key or index used on a mapping or sequence is invalid or not found.
    The exceptions raised are
        1. KeyError
        2. IndexError

```python
#Example - IndexError

try:
    a = [1,2,3,4]
    print(a[10])
except LookupError:
    print("Index out of range")
else:
    print("success")
```

```python
#Example - KeyError

d = {'a':1,'b':2}
print(d['c'])
```
