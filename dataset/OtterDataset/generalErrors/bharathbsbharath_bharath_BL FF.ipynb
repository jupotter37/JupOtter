```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
import pandas as pd
def pick(n):
    result=[]
    for i in range(1,n+1):
        result.append(random.choice(['love candy','love soda','dont love candy','dont love soda']))
    return result
result=pick(16)
print(result)
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x=result)
plt.title('Graph')
plt.xlabel('color')
plt.ylabel('range')
plt.show()
```

```python
phrase_counts = pd.Series(result).value_counts()

plt.figure(figsize=(6, 6))
plt.pie(phrase_counts, labels=phrase_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Phrases")
plt.axis('equal')
plt.show()
```
