```python
while True:
    x = input("Insert a number (q to quit)")
    try:
        if x == 'q':
            break
        x = int(x)
        if int(x) == 0:
            print(10 + x)
        else:
            print("Continue...")
        print(10/x)
    except ValueError:
        print("ValueError! Let's try again.")
    except ZeroDivisionError:
        print("ZeroDivisionError! You can't divide 10 by zero")
    except Exception:
        print("Exception! A generic error occurred")
```
