*Как и список, но хранит только уникальные объекты (уникальность определяется по id объекта)*

```python
set()
```

```python
my_set = set([1,2,3,4,5,3,3,21,4,5,6,6]) ## создаём можество
```

```python
my_set ##создалось множество (упорядоченных чисел)
```

```python
other_set = {3, 4, 5, 2, 5, 6, 8, 8, 9, 6, 5} ## второй способ создания множества - через фигурные скобки {}
```

```python
other_set
```

```python
my_set | other_set ## объединение множеств
```

```python
my_set & other_set ## пересечение множеств
```

*Вычитание множеств:*

```python
my_set - other_set ## попали только те элементы из первого списка, которых нет во втором
```

```python
other_set - my_set ## остались элементы второго скиска, которых нет в первом
```

```python
{1,2,3,4} > {1,2,3,3} ## сравнение множеств
```

```python
21 in my_set ## проверка вхождения элемента в множество
```

*Методы множества:*

```python
my_set
```
