# <a id='toc3_'></a>[Stack](#toc0_)

## <a id='toc3_1_'></a>[What is Stack?](#toc0_)

A `stack` is a `linear data structure` that stores items in a `Last-In/First-Out (LIFO)` or `First-In/Last-Out (FILO)` manner. In stack, a new element is added at one end and an element is removed from that end only. The `insert` and `delete` operations are often called `push` and `pop`.

let's take an example of a `stack of colored plates`. Assume that you have a stack of plates on your table. The first plate you put on the table is at the bottom, so it will be the last one to be used. The last plate you put on the table is at the top, so it will be the first one to be used. This is exactly how a `stack` works.

here is a picture of a `stack`:

![stack](https://cdn.buttercms.com/PuR6MmOQQdqAP6xfh6JO)

Here we start with an `empty stack`. We can `push` items onto the `stack` one by one. Each time we `push` an item, it goes on the `top` of the `stack`. When we `pop` an item off the `stack`, we always `pop` the `top` item. The `stack` is `empty` when there are no items on it.

So, when we `push` an `item` in the `stack` it's always the last `item` and when we `pop` an `item` from the `stack` it's always the last `item` as well. This is why we call it `Last-In/First-Out (LIFO)` or `First-In/Last-Out (FILO)`.

Although `stack` is a `static data structure` that means we have to specify the `size` of the `stack` before using it. But as we are using `arrays` and `linked lists` on `python` to implement `stack` and we can make a `dynamic stack` using `python` `list` and `linked list`.

> Note: `Stack` is a `abstract data type` and `array` and `linked list` are the `data structures` that we use to implement `stack`.


*** Overflow and Underflow

`Overflow` condition occurs when we try to `push` an `item` into a `stack` that is `full`. A `stack` may be `full` because of two reasons:

- Either it is a `fixed size` `stack` and it is `full`.
- Or it is a `dynamic size` `stack` and the `system memory` is `full`.
  
`Underflow` condition occurs when we try to `pop` an `item` from a `stack` that is `empty`. A `stack` may be `empty` because of two reasons:

- Either it is a `fixed size` `stack` and it is `empty`.
- Or it is a `dynamic size` `stack` and the `system memory` is `empty`.

We, will implement a `dynamic size` `stack` using `python` `list` and `linked list` in this article. So, we don't have to worry about `overflow` and `underflow` conditions.

## <a id='toc3_2_'></a>[Stack Operations](#toc0_)

A `stack` is an `abstract data type` that supports the following operations:

- `push`: Adds an `item` to the `stack`. If the `stack` is `full`, then it is said to be an `Overflow condition`.
- `pop`: Removes an `item` from the `stack`. The `items` are `popped` in the `reverse order` in which they are `pushed`. If the `stack` is `empty`, then it is said to be an `Underflow condition`.
- `peek`: Returns the `top` element of the `stack`.
- `isEmpty`: Returns `true` if the `stack` is `empty`, else `false`.
- `isFull`: Returns `true` if the `stack` is `full`, else `false`.(Not mendaotry for `python` implementation)
- `size`: Returns the `size` of the `stack`. (Not mendaotry for `python` implementation)

### <a id='toc3_3_3_'></a>[Peek Operation](#toc0_)

The `peek` operation is also very simple. We will use the `[-1]` index to get the `top` element of the `stack` and `return` it.

````python
class Stack:
    def __init__(self, *items) -> None:
        self.stack = []

        # check if items is not empty
        if items:
            for item in items:
                self.stack.append(item)

    def print_stack(self):
        print(self.stack)

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        return self.stack.pop()
    
    def peek(self):
        return self.stack[-1]
    
new_stack = Stack(1, 2, 3, 4, 5)

print(f"the whole stack:")
new_stack.print_stack()

print(f"the top item: {new_stack.peek()}")
````

### <a id='toc3_3_5_'></a>[Size and Clear Operation ( isFull not required because we are making a dynamic stack)](#toc0_)

### <a id='toc3_4_2_'></a>[Stack Implementation](#toc0_)

`Linked List` is a `dynamic data structure` that means we don't have to specify the `size` of the `linked list` before using it. 

Stack is like a `upside down linked list`. The `top` of the `stack` is the `head` of the `linked list` and the `bottom` of the `stack` is the `tail` of the `linked list`. 

Here, the `head` is the top because we need to be a little clever. If we make the `tail` the `top` then we have to traverse the whole `linked list` to `push` an `item` into the `stack` which has a `time complexity` of `O(n)` but if we make the `head` the `top` then we can `push` an `item` into the `stack` in `O(1)` time.

So, we will make a `Stack` class and initialize it with a `top` attribute and we don't really nead a `bottom` attribute because we are `pushing` and `popping` from the `top` only. And we will also make a `size` attribute to keep track of the `size` of the `stack`.

````python
class Node:
    def __init__(self, data) -> None:
        self.data = data # data of the node
        self.next = None # pointer to the next node

class Stack:
    def __init__(self,value):
        #create a new node
        new_node = Node(value)
        self.top = new_node
        self.height = 1

new_stack = Stack(1)
print(f"new_stack.top.data: {new_stack.top.data}")
````

### <a id='toc3_4_3_'></a>[Push Operation](#toc0_)

Now, let's implement the `push` operation. We will make a `new_node` and assign it to the `top` of the `stack`. Then we will make the `new_node` the `top` of the `stack` and `increment` the `size` of the `stack` by `1`.

It's almost like implementing the `linked list` equivalent of `prepend` method.

````python
class Stack:
    def __init__(self,value):
        #create a new node
        new_node = Node(value)

        # set top to the new node
        self.top = new_node
        self.height = 1

    def push(self, value):
        # create a new node
        new_node = Node(value)

        # set the new node's next to the current top
        new_node.next = self.top

        # set the new node to the top
        self.top = new_node

        # increment the height
        self.height += 1

        return True 
    

new_stack = Stack(1)
print(f"new_stack.top.data: {new_stack.top.data}")
new_stack.push(2)
new_stack.push(3)

print(f"new_stack.top.data after pushing 2 items: {new_stack.top.data}")
        
````

### <a id='toc3_4_4_'></a>[Print Operation](#toc0_)

We will make a `print` method that will print the `stack` in a `list` format. We will make a `current_node` and assign it to the `top` of the `stack`. Then we will loop through the `stack` and print the `data` of each `node`.



````python
class Stack:
    def __init__(self,value):
        #create a new node
        new_node = Node(value)

        # set top to the new node
        self.top = new_node
        self.height = 1

    def push(self, value):
        # create a new node
        new_node = Node(value)

        # if the stack is empty then set the new node to the top
        if self.height == 0:
            self.top = new_node
            self.height += 1
            return True
        
        # set the new node's next to the current top
        new_node.next = self.top

        # set the new node to the top
        self.top = new_node

        # increment the height
        self.height += 1

        return True
    
    def print_stack(self):
        # create a new node to keep track of the current node
        current_node = self.top

        # loop through the stack
        while current_node:
            print(current_node.data)
            current_node = current_node.next

new_stack = Stack(1)
new_stack.push(2)
new_stack.push(3)
new_stack.push(4)
new_stack.push(5)

print(f"the whole stack:")
new_stack.print_stack()
````

> Note: I also made some changes to the `push` operation to check if the `stack` is `empty` or not. If the `stack` is `empty` then we will `push` the `item` as the `top` of the `stack` and `increment` the `size` of the `stack` by `1`. If the `stack` is not `empty` then everything will be the same.

### <a id='toc3_4_5_'></a>[Pop Operation](#toc0_)

The `pop` operation is the `linked list` equivalent of `popfirst` method. We will make a `current_node` and assign it to the `top` of the `stack`. Then we will make the `top` of the `stack` the `next node` of the `current_node` and `decrement` the `size` of the `stack` by `1`. 

We have keep eye on the `size` of the `stack` as well. If the `size` of the `stack` is `0` then we will `return` `None` because there's nothing to `pop` from the `stack`.

````python
class Stack:
    def __init__(self,*values):
        #create a new node
        self.Top = None
        #starting height is 0
        self.height = 0
        
        # check if values is not empty
        if values:
            # loop through the values and push them to the stack
            for value in values:
                self.push(value)

    def push(self, value):
        # create a new node
        new_node = Node(value)

        # if the stack is empty then set the new node to the top
        if self.height == 0:
            self.top = new_node
            self.height += 1
            return True
        
        # set the new node's next to the current top
        new_node.next = self.top

        # set the new node to the top
        self.top = new_node

        # increment the height
        self.height += 1

        return True
    
    def print_stack(self):
        # create a new node to keep track of the current node
        current_node = self.top

        # loop through the stack
        while current_node:
            print(current_node.data)
            current_node = current_node.next

    def pop(self):
        # check if the stack is empty
        if self.height == 0:
            return None

        # get the top item
        top_item = self.top

        # set the top to the next item
        self.top = top_item.next
        # remove the top item connection
        top_item.next = None

        # decrement the height
        self.height -= 1

        return top_item
    

# now we can create a stack with multiple items
new_stack = Stack(1, 2, 3, 4, 5)

print(f"the whole stack:")
new_stack.print_stack()

print(f"popping the top item......")
popped_item = new_stack.pop()

print(f"the whole stack:")
new_stack.print_stack()

print(f"popped item: {popped_item.data}")
````

### <a id='toc3_4_6_'></a>[Peek, isEmpty and Clear Operation](#toc0_)

`Peek` is `top` of the `stack`, `isEmpty` returns `true` if the `stack` is `empty` and `false` if the `stack` is not `empty` and `clear` clears the `stack`.

````python
class Stack:
    def __init__(self,*values):
        #create a new node
        self.Top = None
        #starting height is 0
        self.height = 0
        
        # check if values is not empty
        if values:
            # loop through the values and push them to the stack
            for value in values:
                self.push(value)        

    def push(self, value):
        # create a new node
        new_node = Node(value)

        # if the stack is empty then set the new node to the top
        if self.height == 0:
            self.top = new_node
            self.height += 1
            return True
        
        # set the new node's next to the current top
        new_node.next = self.top

        # set the new node to the top
        self.top = new_node

        # increment the height
        self.height += 1

        return True
    
    def print_stack(self):
        # create a new node to keep track of the current node
        current_node = self.top

        if self.height == 0:
            return None
        
        # loop through the stack
        while current_node:
            print(current_node.data)
            current_node = current_node.next

    def pop(self):
        # check if the stack is empty
        if self.height == 0:
            return None

        # get the top item
        top_item = self.top

        # set the top to the next item
        self.top = top_item.next
        # remove the top item connection
        top_item.next = None

        # decrement the height
        self.height -= 1

        return top_item
    

    def peek(self):
        # check if the stack is empty
        if self.height == 0:
            return None
        
        return self.top
    
    def is_empty(self):
        # if stack has items, return False
        if self.height:
            return False
        
        return True
    
    def clear(self):
        self.top = None
        self.height = 0
    

# now we can create a stack with multiple items
new_stack = Stack(12, 23, 34, 45, 56)

print(f"the whole stack:")
new_stack.print_stack()

print(f"the top item: {new_stack.peek().data}")

print(f"is stack empty? {new_stack.is_empty()}")

print(f"clearing the stack......")
new_stack.clear()

print(f"the whole stack:")
new_stack.print_stack()

print(f"is stack empty? {new_stack.is_empty()}")
````

## <a id='toc4_1_'></a>[What is Queue?](#toc0_)

A `queue` is a `linear data structure` that stores items in a `First-In/First-Out (FIFO)` or `Last-In/Last-Out (LILO)` manner. In `queue`, a new `element` is added at one `end` and an `element` is removed from the `other end` only. The `insert` and `delete` operations are often called `enqueue` and `dequeue`.

let's take an example of a `queue of people`. Assume that you have a queue of people waiting in line to buy tickets. The first person you put in the queue is at the front, so it will be the first one to buy the ticket. The last person you put in the queue is at the back, so it will be the last one to buy the ticket. This is exactly how a `queue` works.

here is a picture of a `queue`:

![queue](https://s3.ap-south-1.amazonaws.com/s3.studytonight.com/curious/uploads/pictures/1582608897-71449.png)

Here we start with an `empty queue`. We can `enqueue` items into the `queue` one by one. Each time we `enqueue` an item, it goes at the `back` of the `queue`. When we `dequeue` an item from the `queue`, we always `dequeue` the `front` item. The `queue` is `empty` when there are no items in it.

So, when we `enqueue` an `item` in the `queue` it's always the last `item` and when we `dequeue` an `item` from the `queue` it's always the first `item` as well. This is why we call it `First-In/First-Out (FIFO)` or `Last-In/Last-Out (LILO)`.

### <a id='toc5_1_2_'></a>[Solution](#toc0_)

We can solve this problem using `stack`. We will make a `stack` and `push` the `opening brackets` into the `stack`. Then we will `pop` the `opening bracket` from the `stack` when we encounter a `closing bracket`. If the `closing bracket` is not the `pair` of the `opening bracket` then we will `return` `false`. If the `stack` is `empty` at the end then we will `return` `true` else we will `return` `false`.

*** Steps:

- We will make a `stack` and `push` the `opening brackets` into the `stack`.
- We can use a hash map to store the `opening brackets` and their `closing brackets` as `key-value pairs`.

- Then we will `pop` the `opening bracket` from the `stack` when we encounter a `closing bracket`. If the `closing bracket` is not the `pair` of the `opening bracket` then we will `return` `false`.

- If the `stack` is `empty` at the end then we will `return` `true` else we will `return` `false`.


*** Implementation

```python

class Solution:
    def isValid(self, s: str) -> bool:
        stack = [] # making a stack
        pairs = {
            '(': ')',
            '{': '}',
            '[': ']'
        } # making a hash map to store the opening brackets and their closing brackets as key-value pairs
        
        for char in s: # looping through the string
            if char in pairs: # if the char is in the pairs
                stack.append(char) # then we will push the char into the stack
            elif len(stack) == 0 or char != pairs[stack.pop()]: # if the stack is empty or the char is not the pair of the opening bracket
                return False # then we will return false

        return len(stack) == 0 # if the stack is empty then we will return true else we will return false
```

*** Complexity

- Time Complexity: `O(n)` where `n` is the length of the `string`.

- Space Complexity: `O(n)` where `n` is the length of the `string`.

```python
        Push element x to the back of queue.
        """
        while self.s1: # while the s1 stack is not empty
            self.s2.append(self.s1.pop()) # we will pop all the items from the s1 stack and push them into the s2 stack

        self.s1.append(x) # then we will append the item into the s1 stack

        while self.s2: # while the s2 stack is not empty
            self.s1.append(self.s2.pop()) # we will pop all the items from the s2 stack and push them into the s1 stack



    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        return self.s1.pop() # we will pop the item from the s1 stack

    def peek(self) -> int:
        """
        Get the front element.
        """
        return self.s1[-1] # we will return the top item of the s1 stack

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        return not self.s1 # we will return true if the s1 stack is empty else we will return false
```

*** Complexity

- Time Complexity: `O(n)` where `n` is the length of the `input queue`.
- Space Complexity: `O(n)` where `n` is the length of the `input queue`.

### <a id='toc5_3_2_'></a>[Solution](#toc0_)

SO, you might be thinking what is `Reverse Polish Notation`. Well, it's a way of writing `arithmetic expressions` without using `parentheses`. It's also known as `postfix notation`. 

If you know the `prefix notation` or `infix notation` then you can easily understand `postfix notation` and the problem should be very easy to solve.

SO, how do you write `postfix notation`? Well, you write it like this:

```

2 1 + 3 *

```

Here, `2` and `1` are the `operands` and `+` and `*` are the `operators`. So, the `postfix notation` of `((2 + 1) * 3)` is `2 1 + 3 *`. SO, here's a `rule` for `postfix notation`:

- `operands` are written first and then the `operators` are written.


SO, how do you evaluate `postfix notation`? Well you search for the `frist operator` and `apply` it to the `operands` and then you `repeat` the process until you have only one `operand` left. That's the `result`.

SO, the `postfix notation` of `((2 + 1) * 3)` is `2 1 + 3 *`. Now, let's evaluate it.

- First we will search for the `first operator` which is `+` and `apply` it to the `operands` which are `2` and `1`. So, `2 + 1 = 3`. Now, we have `3 3 *` left.

- Then we will search for the `first operator` which is `*` and `apply` it to the `operands` which are `3` and `3`. So, `3 * 3 = 9`. Now, we have `9` left.

- Now, we have only one `operand` left which is `9`. So, `9` is the `result`.

SO, that's how you evaluate `postfix notation`. Now, let's implement it.

*** Steps:

- We will make a `stack` that will store the `operands`.
- Then we will loop through the `tokens`.
- If the `token` is an `operand` then we will `push` it into the `stack`.
- If the `token` is an `operator` then we will `pop` the `top 2 operands` from the `stack` and `apply` the `operator` to them and `push` the `result` into the `stack`.
- At the end we will `return` the `top operand` of the `stack`.
  
*** Implementation

```python

class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = [] # making a stack

        for token in tokens: # looping through the tokens
            if token in '+-*/': # if the token is an operator
                num2 = stack.pop() # then we will pop the top 2 operands from the stack
                num1 = stack.pop()
                if token == '+': # and apply the operator to them
                    stack.append(num1 + num2) # and push the result into the stack
                elif token == '-':
                    stack.append(num1 - num2)
                elif token == '*':
                    stack.append(num1 * num2)
                elif token == '/':
                    stack.append(int(num1 / num2))
            else: # if the token is an operand
                stack.append(int(token)) # then we will push it into the stack

        return stack[-1] # at the end we will return the top operand of the stack
```

*** Complexity

- Time Complexity: `O(n)` where `n` is the length of the `tokens`.
- Space Complexity: `O(n)` where `n` is the length of the `tokens`.

### <a id='toc5_4_2_'></a>[Solution](#toc0_)

*** Steps:

- We will make a `stack` that will store the `items`.
- We will make a `min` variable that will store the `minimum element`.
- When an `item` is `pushed` into the `stack` we will check if the `item` is the `minimum element` or not. 
- If the `item` is the `minimum element` then we will `push` the `item` into the `stack` and `update` the `min` variable.
- If the `item` is not the `minimum element` then we will `push` the `item` into the `stack` without `updating` the `min` variable.
- When an `item` is `popped` from the `stack` we will check if the `item` is the `minimum element` or not.
- If the `item` is the `minimum element` then we will `pop` the `item` from the `stack` and `update` the `min` variable.
- If the `item` is not the `minimum element` then we will `pop` the `item` from the `stack` without `updating` the `min` variable.
- When we call `getMin` method we will `return` the `min` variable.

*** Implementation

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.minn = float("inf") # making a min variable that will store the minimum element

    def push(self, val: int) -> None:
        if val <= self.minn:
            self.stack.append(self.minn) # if the item is the minimum element then we will push the item into the stack and update the min variable
            self.minn = val
        self.stack.append(val)

    def pop(self) -> None:
        if self.stack.pop() == self.minn: # if the item is the minimum element then we will pop the item from the stack and update the min variable
            self.minn = self.stack.pop()

    def top(self) -> int:
        return self.stack[-1] # we will return the top item of the stack

    def getMin(self) -> int:
        return self.minn # we will return the min variable

```

*** Complexity

- Time Complexity: `O(1)` for `push`, `pop`, `top` and `getMin` methods.
- Space Complexity: `O(n)` where `n` is the length of the `stack`.

### <a id='toc5_5_2_'></a>[Solution](#toc0_)

This is porblem that needs a little unconventional thinking. 

If you look at the `rainwatertrap` image you will see that the `rainwater` is trapped between the `bars`. So, we need to find the `bars` that are `trapping` the `rainwater`.

And also the `rainwater` is trapped between the `bars` so we need to find the `minimum height` of the `bars` that are `trapping` the `rainwater`.

And also we need to find the minimum height of the `bars` that are `trapping` the `rainwater` from the `left` and the `right` side of the `bar`.

And lastly we need to find the `difference` between the `minimum height` of the `bars` that are `trapping` the `rainwater` from the `left` and the `right` side of the `bar` and the `height` of the `bar`.

Now the pattern is that if the height of the `position` is `less` than the `minimum height` of the `bars` that are `trapping` the `rainwater` from the `left` and the `right` side of the `bar` then the `difference` between the `minimum height` of the `bars` that are `trapping` the `rainwater` from the `left` and the `right` side of the `bar` and the `height` of the `bar` will be the `amount of rainwater` that is `trapped` in that `position`.

Now, I'm bieng honest here, the only good thing about this problem is that it wants you to return how much `rainwater` is `trapped` and not the `positions` of the `bars` that are `trapping` the `rainwater`. If it wanted you to return the `positions` of the `bars` that are `trapping` the `rainwater` then it would be a pain in the butt to solve this problem.

Now, let's implement it.

*** Steps:

- We will make two `maxleft` and `maxright` arrays that will store the `maximum height` of the `bars` that are `trapping` the `rainwater` from the `left` and the `right` side of the `bar`.
- Then we will loop through the `height` array and find the `maximum height` of the `bars` using `min(maxleft[i], maxright[i])` and `subtract` the `height` of the `bar` from it.

*** Implementation

```python

class Solution:
    def trap(self, height: List[int]) -> int:
        n = len(height)
        maxleft = [0] * n # making a maxleft array that will store the maximum height of the bars that are trapping the rainwater from the left side of the bar
        maxright = [0] * n # making a maxright array that will store the maximum height of the bars that are trapping the rainwater from the right side of the bar

        maxleft[0] = height[0] # we will initialize the first item of the maxleft array with the first item of the height array
        maxright[-1] = height[-1] # we will initialize the last item of the maxright array with the last item of the height array

        for i in range(1, n): # looping through the height array
            maxleft[i] = max(maxleft[i - 1], height[i]) # we will find the maximum height of the bars using min(maxleft[i], maxright[i]) and subtract the height of the bar from it

        for i in range(n - 2, -1, -1): # looping through the height array
            maxright[i] = max(maxright[i + 1], height[i]) # we will find the maximum height of the bars using min(maxleft[i], maxright[i]) and subtract the height of the bar from it


        ans = 0 # making a ans variable that will store the amount of rainwater that is trapped

        for i in range(n): # looping through the height array
            ans += min(maxleft[i], maxright[i]) - height[i] # we will find the maximum height of the bars using min(maxleft[i], maxright[i]) and subtract the height of the bar from it

        return ans # at the end we will return the ans variable

```

*** Complexity

- Time Complexity: `O(n)` where `n` is the length of the `height` array.

This one took me a lot of time to understand. And if understand this problem somehow it becomes more confusing. I don't know why this problem is in the `grind 75` list. But we will do it anyway.

It's like the `rainwater` problem. We need to find the `bars` that are `trapping` the `rainwater`. But this time we need to find the `area` of the `rectangle` that is formed by the `bars` that are `trapping` the `rainwater`.

The solution is nothing like the `rainwater` problem. 

So, a `bar` can be extended to the `left` and the `right` until we find a `bar` that is `less` than the `current bar`. And the `area` of the `rectangle` will be the `height` of the `current bar` multiplied by the `distance` between the `current bar` and the `bar` that is `less` than the `current bar`.

For the limitation of this article I can't explain the problem properly. I hope you can understand the solution better.

```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        maxArea  = 0 # making a maxArea variable that will store the maximum area of the rectangle
        stack = [] # making a stack

        for idx, height in enumerate(heights): # looping through the heights array
            start = idx # we will initialize the start variable with the idx
            while stack and stack[-1][1] > height: # while the stack is not empty and the top item of the stack is greater than the height
                idx, h = stack.pop() # we will pop the top item of the stack
                maxArea = max(maxArea, h * (start - idx)) # we will find the area of the rectangle and update the maxArea variable

            stack.append((start, height)) # we will append the (start, height) tuple into the stack
```

```python
class Solution:
    def calculate(self,s):
        num = 0
        ans = 0
        sign = 1
        stack = [sign]

        for c in s:
            if c.isdigit():
                num = num * 10 + int(c) # we will get the number

            elif c == "(":
                stack.append(sign) # we will push the previous sign to the stack because the new sign will be the sign of the expression in the parenthesis
            elif c == ")":
                stack.pop() # we will pop the top item of the stack
            elif c in "+-":
                ans += sign * num # we will add the oparends according to the sign
                sign = (1 if c == "+" else -1) * stack[-1] # we will update the sign
                num = 0 # we will reset the number

        ans += sign * num # we will add the last number to the stack

        return ans # at the end we will return the ans variable

```
