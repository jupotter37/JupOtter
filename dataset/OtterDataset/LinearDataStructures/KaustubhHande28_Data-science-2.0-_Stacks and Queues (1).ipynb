# Stacks:

**Definition:** A stack is a linear data structure that follows the LIFO (Last In First Out) prinical. So, the last element inserted in the data structure is removed first.

**Why and When do we use Stack?**
* Stacks are simple data structures that allow us to store and retrieve data sequentially.

* Imagine you’re a developer and you are working on a brand new word processor. You need to create an undo feature — allowing users to backtrack their actions until the beginning of the session. A stack is an ideal fit for this scenario. We can record every action of the user by pushing it to the stack. When the user wants to undo an action they can pop accordingly from the stack.


### Real life examples:

1. **Undo/Redo Functionality:**
Undo and redo functionalities are present in software applications, from text editors to graphic design software. Stacks are employed to maintain a history of user actions, allowing for easy reversibility and exploration of past changes. When an action is performed, it's pushed onto the undo stack, and if the user wishes to undo it, the action is popped from the undo stack and pushed onto the redo stack, ensuring a seamless way to navigate through changes and correct mistakes.

2. **Balanced Parentheses:**
Check if brackets { } ( ) [ ] are used correctly in an expression. Add opening brackets to the stack and remove them when you see the corresponding closing bracket. If everything is balanced, the stack should be empty at the end.


3. **Expression Evaluation (Infix, Postfix, Prefix):**
We can use a stack to evaluate mathematical expressions written in different notations:
 * Infix notation: The usual way we write expressions, with operators (+, -, *, /) between operands (numbers). (e.g., 2 + 3 * 4)
 * Postfix notation: Operators come after their operands. (e.g., 2 3 4 *)
 * Prefix notation: Operators come before their operands. (e.g., * 2 3 4)
The stack helps us process these expressions in the correct order based on the operator precedence rules. We can "push" operands and operators onto the stack and then "pop" them in a specific order depending on the notation (infix, postfix, prefix) to perform the calculations and get the final result.

### Basic stack operations:

```python
# Implenatation of stack using list

class Stack:

    def __init__(self):
        """
        init is a reserved method in Python that initialises the attributes/variables of a class.
        It is called automatically once an object of class Stack is created.
        """
        self.items=[]

    def push(self, item):
        """
        push will add an item to the top of the stack. It takes the item to be pushed as an input parameter
        and modifies the stack. Returns none.
        """
        # There is no overflow condition here. why? we have created an empty list which
        # has dynamic size. If we had created a list of fixed size, then we would have to
        # check for overflow condition before pushing into the stack.
        self.items.append(item)

    def pop(self):
        """
        removes the top most item from the stack and returns the item. The original stack is modified
        """
        #underflow condition.
        if self.isEmpty():
            return "The stack is empty"

        return self.items.pop()

    def isEmpty(self):
        """
        returns True if the stack is empty, False otherwise. Needs no input parameter
        """
        return self.items==[]

    def peek(self):
        """
        returns the top most item of the stack. No parameter is required. The original stack is not modified
        """
        return self.items[len(self.items)-1]

    def size(self):
        """
        returns the number of items in the stack. No parameter is required.
        """
        return len(self.items)
```

### Balanced paranthesis

Write an algorithm which returns "True" if a group of paranthesis passed are balanced. If not, it returns "False". Note that the paranthesis can be of simple brackets '()' type only

Sample input:
(()()()()) \
Sample Output:
True

Sample input:
(()((())())) \
Sample Output:
True

Sample input:
((((((()) \
Sample Output:
False

As you process symbols from left to right, the most recent opening parenthesis must match the next closing symbol (see Figure 4). Also, the first opening symbol processed may have to wait until the very last symbol for its match. Closing symbols match opening symbols in the reverse order of their appearance; they match from the inside out. This is a clue that stacks can be used to solve the problem.

1. Starting with an empty stack, process the parenthesis strings from left to right.
2. If a symbol is an opening parenthesis, push it on the stack as a signal that a corresponding closing symbol needs to appear later.
3. If, on the other hand, a symbol is a closing parenthesis, pop the stack. As long as it is possible to pop the stack to match every closing symbol, the parentheses remain balanced.
4. If at any time there is no opening symbol on the stack to match a closing symbol, the string is not balanced properly.
5. At the end of the string, when all symbols have been processed, the stack should be empty.

```python
def balanced_paranthesis(string):
    """
    returns True is the paranthesis are balanced, else False
    """
    s=Stack()
    index=0
    balanced=True

    # Parse through the given string as long as the interator(index) reaches the end of string and the brackets are balanced.
    # As soon as the code encounters the first non-balanced bracket, it should come out of the loop and return False
    while (index<len(string) and balanced):
        if string[index]=='(':
            s.push(string[index])

        else:

            #When the number of closing brackets > opening brackets, the stack can go empty
            if s.isEmpty():
                balanced=False

            #When the stack s is not empty and the string[index] != '(' in a particular iteration
            else:
                s.pop()

        #increment the index(iterator) on each iteration
        index+=1

    # At any point if the stack becomes empty (balance=False), control will pass to this if statement
    #and the function will return false. Another possibility of imbalanced paranthesis is
    #when stack doesn't become empty until last iteration. Likely to happen when the no. of opening braces > closing braces
    # Hence we need to check for both the conditions for the function to return "True"
    if balanced and s.isEmpty():
        return True
    else:
        return False
```

```python
string='((()))'
print(balanced_paranthesis(string))
print('---')
print(balanced_paranthesis('((()))'))
print(balanced_paranthesis('(()'))
```

```python
# Base line model

def balanced_paranthesis(string):
    """
    returns True is the paranthesis are balanced, else False
    """

    s = Stack()
    index = 0

    while index < len(string):
        if string[index] == '(':
            s.push(string[index])
        elif string[index] == ')':
            if s.isEmpty():
                return False
            s.pop()
        index += 1

    return s.isEmpty()
```

### Valid Parenthesis

Write an algorithm that returns 'True' if the input given is a balanced/valid parenthesis and 'False' incase of unbalanced paranthesis.

Sample Input: s = "()" \
Sample Output: True

Sample Input: s = "()[]{}" \
Sample Output:True

Sample Input: s = "{[]}" \
Sample Output: True

Sample Input: s = "([)]" \
Sample Output: False

```python
def valid_par(string):
    """
    returns True is the paranthesis are balanced, else False
    """
    open={'(','{','['} #a set that has the opening bracket of each type

    close=  {')': '(',  #a dictionary that has the end bracket as a key and the corrosponding open bracket as value
             '}': '{',
             ']': '['}

    stack=[]

    for char in string:
        if char in open:
            stack.append(char)
        elif len(stack)>0 and close[char]==stack[-1]: #notice the order of the conditions
            stack.pop()

        else:
            return False

    return stack==[]
```

```python
s = "()[]{}"
print(valid_par(s))
```

```python
s = "([)]"
print(valid_par(s))
```

* **Underflow:** Trying to pop/peek an element from an empty stack. For underflow, you saw that we framed a condition wherein a pop is not allowed from an empty stack.

* **Overflow:** Trying to push an element into a stack that is already at its maximum capacity. A stack overflow occurs when you have used up more memory for a stack than your program was supposed to. ## Limitations of Stacks:

They have limitations that make them less suitable for tasks requiring access to arbitrary elements:

* No Random Access:
<br>
Unlike arrays, which allow constant-time access to elements based on their index, and linked lists, which can be traversed to access any element by following pointers, stacks only allow access to the most recently added element. You cannot directly access elements in the middle or at the bottom of the stack without first removing the elements on top.


* Limited Data Manipulation:
<br>
 Stacks are designed for sequential data manipulation, where elements are added and removed in a specific order. They are not optimized for tasks that involve rearranging or searching for elements within the stack, which are operations more efficiently handled by arrays or linked lists.

```python
# Implenatation of queue using lists

class Queue:
    def __init__(self):
        self.items=[]

    def enqueue(self, item):
        """
        enqueue will add an item to the rear of the queue. Its needs an item as input parameter
        and modifies the queue. returns nothing.
        """
        self.items.insert(0,item)

    def dequeue(self):
        """
        removes the front item from the queue and returns the item. The queue is modified
        """
        #underflow condition
        if self.isEmpty():
            return("Queue is empty")
        return self.items.pop()

    def isEmpty(self):
        """
        returns True if the queue is empty, False otherwise.
        """
        return self.items==[]

    def size(self):
        """
        returns the number of items in the queue. No parameters needed. returns an integer.
        """
        return len(self.items)
```

### Game of Hot Potato simulation

Did you know that one of our favourite childhood games - **`Musical Chairs** can be simulated using the queue operations that you just learnt? \

Here, we'll try out a popular variant of this game, called the **Hot Potato Problem**.In this game, a group of people line up in a circle, and once the music starts, players pass an object(say a ball or a potato) to the next person and keep doing that as long as the music is still playing. The moment the music stops, whoever has the ball in their hand will be removed from the circle, and the next round of the game starts. The loop continues until only one player is left and that person is declared the winner.

To implement the logic using a queue, we need to implement the following steps

- Store the given list of players in a queue. Assume the object is with the player at the top of the queue.
- The number of passes, after which a player is eliminated is taken as an input. *For simplicity, we'll assume that this number is **constant**.*
- Perform the necessary enqueue and dequeue operations based on the number of passes to denote each round of elimination where a player is removed from the queue and the object is passed to the next person who'll start the subsequent round. *The last person remaining will be declared as winner.*

Let's look at the following testcase to understand it better

`Sample input:` \['Smith', 'Garry', 'Michael', 'Julian', 'Robert'\], 3 \
`Sample Output:`Smith

*Explanation:* A list of players and the number of passes after which a player should be eliminated is given as input. The output will be the final winner of the game.

In test case 1, We start by enqueuing the players list into a queue. Since queue follows FIFO principle, the final queue will look like: \
`['Robert', 'Julian', 'Michael', 'Garry', 'Smith'] `

The number of passes after which a player should be eliminated is 3. The game starts with Smith having the ball who passes it to the person next to him,that is, Garry. Then Smith leaves the front and joins back in the rear of queue. Now, the queue will look like: \
`['Smith','Robert', 'Julian', 'Michael', 'Garry']`

Now, Garry passes the ball to Michael, leave from the front and join at the rear end. Now, the queue will look like: \
`['Garry', 'Smith','Robert', 'Julian', 'Michael']`

Now, in the third pass, Michael passes the ball to Julian, leave from the front and join at the rear end. Now, the queue will look like: \
`['Michael','Garry', 'Smith','Robert', 'Julian']`

At the end of third pass, Julian, who is in the front has the ball. Hence, he will be elimated from the game and the queue now becomes:\
`['Michael','Garry', 'Smith','Robert']`

Now, the second round starts with Robert having the ball and passing it to the next player. The ball keeps getting passed 3 times in each round until a player is eliminated. After each round of elimination, the player next to the eliminated person will start the next round of the game.\

Here's a summary of the different rounds and the player that get eliminated in each round.
**' -> '** denotes the object being passed from one player to other.

`Round 1:` Smith -> Garry -> Michael -> Julian. Now, Julian has the ball and he will be eliminated.  

`Round 2:` Robert -> Smith -> Garry -> Michael. Now, Michael has the ball and will be eliminated.  

`Round 3:` Robert -> Smith -> Garry -> Robert. Now, Michael has the ball and he will be eliminated.

`Round 4:` Smith -> Garry -> Smith ->Garry. Now, Garry has the ball and he will be eliminated.

Only Smith is left now and thus he is declared as the winner.

```python
def hot_potato(players,passes):

    # Create an object q of type Queue to use the fuctions that are present in the class queue to write our program
    q=Queue()

    """
    returns the name of the winner in a Hot Potato Game
    """

    #First create a queue of all the players
    for player in players:
        q.enqueue(player)


    # As long as more than one player remains there will be a loop of rounds. Because when the size of queue becomes one,
    #that single player becomes the winner
    while q.size()>1:

    # For num number of passes, remove(dequeue) a player from the front and add him/her to the rear(enqueue) of queue
        for i in range(passes):
            player_passing=q.dequeue()
            q.enqueue(player_passing)

    # after 'num' number of passes. remove the player who is at the front of q using q.dequeue.
        q.dequeue()

    # The control will go back to the while statement now and the next round of game(for loop) starts.
    return q.dequeue()
```

### Limitation:
* Limited Access:
<br>
Queues follow the First-In-First-Out (FIFO) principle, meaning the first element added to the queue is the first one to be removed. While this is advantageous for certain scenarios like task scheduling or managing resources, it restricts access to elements other than the front of the queue. Accessing elements in the middle or end of the queue requires dequeuing all preceding elements, which can be inefficient.


* No Random Access:
<br>
Similar to stacks, queues do not support random access to elements. You can only access and remove elements from the front of the queue. If you need to access elements at arbitrary positions within the queue, a different data structure like an array or linked list would be more suitable.


* Limited Use Cases:
<br>
Queues are most commonly used in scenarios where the order of processing matters and where elements need to be processed in the order they were added. While they excel in tasks like job scheduling, process management, or breadth-first search algorithms, they may not be appropriate for tasks requiring more flexible access patterns or data manipulation.


* Inefficient for Certain Operations:
<br>
Some operations, such as searching for an element within the queue or inserting an element at a specific position, can be inefficient with queues. Due to their sequential nature, searching typically requires dequeuing elements until the desired element is found, and inserting elements at specific positions requires dequeuing and enqueuing elements to maintain the correct order.


* Fixed Size Limitation:
<br>
In implementations of queues using arrays, there may be a fixed size limitation due to the static nature of arrays. Once the queue reaches its maximum capacity, it cannot accommodate additional elements unless it is resized, which can be a costly operation.
