## Time and Space Complexity

#### What is it?

Time and space complexity is the measure of how much time a given action(function) will take to solve a problem. In the same fashion, we determine how much a given data structure will need in terms of memory allocation. A problem can have multiple solutions and finding the optimal solution for the problem needs to be analyzed in time and space.

#### How do we measure Time and Space Complexity?

In order to measure time and space complexity we use Asymptotic analysis. The reason for this is because we need a way to measure different algorithms (functions) based on the size of their inputs in a mathmatical way. For example, we could have a function that is computed as f(n) and another that is g(n^2). All things around the function staying constant, the only thing that changes is the size of the input. ## Stacks and Queues

** Stacks ** as the name suggests is a data structure that allows for data to follow the Last In First Out priciple(LIFO). Think of a stack of pancakes for example. To get the first pancake you would  start with the top and go down.

##### Searching through a stack will be Linear Time O(n) - Constant Space O(1)
##### Selecting the last item will be done in Constant Time O(1) - Constant Space O(1)
##### Adding to the stack should take Constant Time O(1) - Constant Space O(1)

** Queues ** are similar but in this case follow the First In First Out principle(FIFO). Think of this as a line in a black friday sale. The first person camped out for the big screen tv is the first to get it.

##### Searching through a queue will be Linear Time O(n) - Constant Space O(1)
##### Selecting the first item will be done in Constant Time O(1) - Constant Space O(1)
##### Adding to the queue should take Constant Time O(1) - Constant Space O(1)

```python
from collections import deque
#python list version of stack
stack = []
stack.append('pancake1')
stack.append('pancake2')
print(stack)
stack.append('pancake3')
print(stack)
stack.append('packcake4')
print(stack)

stack.pop()
print(stack)
stack.pop()
print(stack)
```

# Sum of Parts Time Compelxity Exercise

```python
# sum of parts
# https://www.codewars.com/kata/5ce399e0047a45001c853c2b
ls = [0, 1, 3, 6, 10]    
def parts_sums(ls):
    sum_list = []
    while len(ls) > 0:
        for num in ls:
            sum_list.append(sum(ls))
            num = ls.pop(0)
    if len(ls) == 0:
        sum_list.append(0)
    return sum_list
```

```python
ls = [0, 1, 3, 6, 10]
def parts_sums(ls):
    result = []
    total_sum = sum(ls)

    for num in ls:
        result.append(total_sum)
        total_sum -= num

    result.append(total_sum)  # Add the final sum of an empty list

    return result
```

## Linked List (Data Structure)

A linked list is created by using the node class. We create a Node object and create another class to use this node object. We pass the appropriate values thorugh the node object to point the to the next data elements.

There are some advantages and disadvantages with this data structure. **Advantages** Linked Lists can save memory because they can be flexibile with memory management which saves memory. **Disadvantages** Finding or adding to the list requires traversing the entire list.

```python
# 2 components of the linked list
# the Node Class and a node object

# linked list class that us tie our node objects together
#Node Class
class Node():
    def __init__(self, value):
        self.value = value
        self.next = None
        
class LinkedList():
    def __init__(self):
        self.head = None
        
#push_on method place an item at the beginning of the list
    def push_on(self, new_value):
        new_node = Node(new_value)
        new_node.next = self.head
        self.head = new_node
        
#insert_after which will insert a node after another node in the linked list
    def insert_after(self, prev_node, new_value):
        #check if previous node exists
        if prev_node is None:
            print("the given prev_node must not be empty")
            return
        #if the prev_node is not empty
        new_node = Node(new_value)
        
        #update the new node's next to be the previous node's next
        new_node.next = prev_node.next
        
        #update prev_node next to be new_node
        pre_node.next = new_node
        
#append add a new node to the end of the list
    def append(self, new_value):
        #create new node object
        new_node = Node(new_value)
        
        #check if the linked list is empty
        #if it is make the new node the head of the linked list
        if self.head is None:
            self.head = new_node
            
        #if not empty, traverse the linked lst and append to the end
        last = self.head
        
        #change last variable to the new value and continuosly check for an empty list
        while last.next:
            last = last.next
        #change the current last node (trail) to point to the new node object
        last.next = new_node
        
#traverse move through the linked list
    def traverse(self):
        temp = self.head
        #while temp is NOT NONE --> kepp on truckin through the list until it is none
        while temp:
            print(temp.value)
            temp = temp.next
        print("End of your list!")



#ex.
#my_linked.push_on('Sunday')
```

```python
# Coding Wars Problem 1
# https://www.codewars.com/kata/56f3a1e899b386da78000732/solutions/python
# 7 kyu
# Parts of a list

def partlist(arr):
    if len(arr) < 2:
        return ("Invalid input")
    else:         
        answer_list = []
        for i in range (1,len(arr)):
            first_part = " ".join(arr[:i])
            second_part = " ".join(arr[i:])
            answer_list.append((first_part, second_part))
        return answer_list
```

```python
# Coding Wars Problem 2
# https://www.codewars.com/kata/54da5a58ea159efa38000836/solutions/python
# 6 kyu
# Find the odd int

def find_it(arr):
    freq_map = {}
    for integer in arr:
        if integer in freq_map:
            freq_map[integer] += 1
        else:
            freq_map[integer] = 1
    
    for integer, frequency in freq_map.items():
        if frequency % 2 != 0:
            return integer
```
