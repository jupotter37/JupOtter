```python
class Stack:
    def __init__(self) -> None:
        self.items = []
    
    def push(self, data):
        self.items.append(data)

    def pop(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
    def is_empty(self):
        return len(self.items) == 0
    
    def peek(self):
        return self.items[-1]
```

```python
# LinkedList를 이용해 Stack class 만드는 방법
class Node:
    def __init__(self) -> None:
        self.data = data
        self.next = None
    
class Stack:
    def __init__(self) -> None:
        self.head = None

    def push(self, data):
        '''
            head를 계속해서 뒤쪽으로 미룬다
        '''
        node = Node(data)
        if self.head is None:
            self.head = node
        else:
            node.next = self.head
            self.head = node
        
    def pop(self):
        if self.head is None:
            raise IndexError('pop from empty stack')
        poppednode = self.head
        self.head = self.head.next 
        return poppednode.data
```

```python
# 가장 작은 값을 찾는 방법
class MinStack():
    def __init__(self) -> None:
        self.main = [] # main stack
        self.min = [] # 가장 작은 요소를 추적
    
    def push(self, n):
        if len(self.main) == 0:
            self.min.append(n)
        elif n <= self.min[-1]:
            self.min.append(n)
        else:
            self.min.append(self.min[-1])
        self.main.append(n)

    def pop(self):
        self.min.pop()
        return self.main.pop()
    
    def get_min(self):
        return self.min[-1]

min_stack = MinStack()
min_stack.push(10)
print(min_stack.main)
print(min_stack.min)
min_stack.push(15)
print(min_stack.main)
print(min_stack.min)
```

```python
# 내부 링크드 리스트를 사용해 Queue class 만들기
class Node:
    def __init__(self, data, next=None) -> None:
        self.data=data
        self.next=next
    
class Queue:
    def __init__(self) -> None:
        self.front=None
        self.rear=None
        self._size = 0
    
    def enqueue(self, item):
        self._size += 1
        node =Node(item)
        if self.rear is None:
            self.front = node 
            self.rear = node 
        else: 
            self.rear.next = node 
            self.rear = node
    
    def dequeue(self):
        if self.front is None:
            raise IndexError('pop from empty queue')
        self._size-=1
        temp=self.front 
        self.front=self.front.next 
        if self.front is None:
            self.rear = None 
        return temp.data
    
    def size(self):
        return self._size
```

```python
# making Queue with two stacks
class Queue:
    def __init__(self) -> None:
        # two stacks
        self.s1=[]
        self.s2=[]
    
    def enqueue(self, item):
        # s1내용을 s2로 옮김
        while len(self.s1) != 0:
            self.s2.append(self.s1.pop())
        # 추가할 내용을 s1에 넣음, 이 내용이 맨 앞으로 넣고, 원래 내용을 다시 추가해야 가장 첫번째 요소를 먼저 추출할 수 잇어서.
        self.s1.append(item)
        # s2 내용을 다시 s1으로 옮김
        while len(self.s2) != 0:
            self.s1.append(self.s2.pop())

    def dequeue(self):
        if len(self.s1) == 0:
            raise Exception('Cannot pop from empty queue')
        return self.s1.pop()
```

```python
class BinaryTree:
    def __init__(self, value) -> None:
        self.key = value
        self.left_child = None 
        self.right_child = None 

    def insert_left(self, value):
        if self.left_child == None:
            self.left_child = BinaryTree(value)
        else:
            bin_tree = BinaryTree(value)
            bin_tree.left_child = self.left_child
            self.left_child = bin_tree 
    
    def insert_right(self, value):
        if self.right_child == None:
            self.right_child = BinaryTree(value)
        else:
            bin_tree = BinaryTree(value)
            bin_tree.right_child = self.right_child
            self.right_child = bin_tree
        
    '''탐색하려는 데이터 n을 매개변수로 받는 함수'''
    def breadth_first_search(self, n): # data : n 
        current = [self] # 현재 탐색 중인 레벨의 노드들 , self=> 레벨0부터 시작.
        next = [] # 다음 레벨의 노드들 
        while current:
            for node in current: # 모든 노드 순회, current 내용을 순회하는 것 
                if node.key == n:
                    return True
                if node.left_child:
                    next.append(node.left_child)
                if node.right_child:
                    next.append(node.right_child)
            current = next 
            next = []
        return False 
```

# DFS

```python
# 전위 이동
# 모든 레벨 왼쪽 먼저 끝까지 간뒤 아래 레벨부터 오른쪽으로 간다.
def preorder(tree):
    if tree:
        print(tree.key)
        preorder(tree.left_child)
        preorder(tree.right_child)
```

```python
# 후위 이동 
def postorder(tree):
    if tree:
        postorder(tree.left_child)
        postorder(tree.right_child)
        print(tree.key)
```

```python
# 중위 이동
def inorder(tree):
    if tree:
        inorder(tree.left_child)
        print(tree.key)
        inorder(tree.right_child)
```

# Reversing Binary Tree
BFS를 이용하여 뒤집기


```python
class BinaryTree:
    def __init__(self, value) -> None:
        self.key = value
        self.left_child = None 
        self.right_child = None 
    
    def insert_left(self, value):
        if self.left_child == None:
            self.left_child = BinaryTree(value)
        else:
            bin_tree = BinaryTree(value)
            bin_tree.left_child = self.left_child 
            self.left_child = bin_tree 

    def insert_right(self, value):
        if self.right_child == None:
            self.right_child = BinaryTree(value)
        else:
            bin_tree = BinaryTree(value)
            bin_tree.right_child = self.right_child 
            self.right_child = bin_tree 
        
    def invert(self):
        current = [self]
        next = []
        while current:
            for node in current:
                if node.left_child:
                    next.append(node.left_child)
                if node.right_child:
                    next.append(node.right_child)
                tmp = node.left_child
                node.left_child = node.right_child
                node.right_child = tmp
            current = next 
            next = []
```
