## Implement a Queue

```python
class SimpleQueue:
    def __init__(self):
        self.queue = list()
    
    # append new data to END of list 
    def enqueue(self, data):
        self.queue.append(data)
    
    # remove new data from START of list
    def dequeue(self):
        if len(self.queue) > 0:
            return self.queue.pop(0)  
        raise IndexError("Queue is empty")
    
    # Peek Queue: PRINT at the FIRST element in Queue 
    def peek(self):
        if len(self.queue) > 0:
            return self.queue[0]
        raise IndexError("No data in Queue")  # allows to handle cases where 
        
    def view_queue(self):
        if len(self.queue) > 0:
            for item in self.queue:
                print(item)
        else:
            raise IndexError("Queue is Empty")
    
songs = SimpleQueue()

# Add items to queue
songs.enqueue('flaca')

songs.enqueue('Arurian Dance')

songs.enqueue('Lolo Felix')

# Peek Queue
print("Next Song in Queue: ", songs.peek())

songs.dequeue()
print("Next Song in Queue: " , songs.peek())

# View Current Queue
print("")
print("Current Queue:")
songs.view_queue()


# Get Queue 2:
print("")
print("Currrent Queue")
print(songs.queue)
```

## Implement a Stack

```python
class SimpleStack:
    def __init__(self):
        self.stack = list()

    def push(self, data):
        self.stack.append(data)

    def remove(self):
        if self.stack > 0:
            self.stack.pop()
        raise IndexError("Stack Is Empty")
    
    def PrintStack(self):
        for item in self.stack[::-1]: # iterate over stack inversely, going from end(top) --> to start (bottom)
            print(item)

pancakes = SimpleStack()

pancakes.push('vanilla')
pancakes.push('chocolate')
pancakes.push('strawbery')

pancakes.PrintStack()
```

## Implement a Singly Linked List

Inserting/Deleting:
- Linked lists have fast INSERTION to the START of the list (pre-pending) because other items don't need to be moved around to make space
- Linked lists have O(n) time for INSERTION at the end (appending) because you have to iterate over all the items in the list
- Linked lists have fast DELETION at the beginning O(1) 
    - O(n) time in worst case scenario (deleting last node) 

conclusion: LINKED LISTS ARE GREAT (better than arrays) AT INSERTING/DELETING DATA AT THE START O(1) 


- Searching a Linked List:  0(n) 



```python
class simpleNode:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

class simpleLinkedList:
    def __init__(self):
        self.head_node = None

    def append(self,data):
        """
        Add values to the end of the List. 
        """
        new_node = simpleNode(data)

        # If there is no head node, set new_node to head node 
        if not self.head_node:
            self.head_node = new_node
            return

        # If there is a head node, iterate over linked list until you reach the last node: last.next = none. Set last.next = new_node
        last = self.head_node
        while last.next:  # when reach the last node, last.next = None, and loop stops
            last = last.next 

        # Append new_node to end of list 
        last.next = new_node 
    
    def pre_pend(self,data):
        """
        Add value to the start of list. 
        1. initialize data as node object
        2. if no head node, set new node to head_node
        3. if there is a head node, set new_node.next attribute to head_node, and re-set head_node to new_node
        """
        new_node = simpleNode(data)
        if not self.head_node:
            self.head_node = new_node
        
        new_node.next = self.head_node
        self.head_node = new_node 

    def delete(self, data):
        """
        1. If linked list is empty: return
        2. If data to delete is the head node data, set new head node to self.head.next  
        3. Iterate over nodes in linked list 
            - when CurrentNode.data = data --> PreviousNode.next = CurrentNode.next 
        """
        if not self.head_node:
            print('list is empty')
            return
                

        if self.head_node.data == data:
            self.head_node = self.head_node.next 
            print('deleted head node')
            return
        
        previous_node = None
        current = self.head_node
        while current and current.data != data: # this loop stops once we reach the end of the list, or when we find a match
            previous_node == current
            current = current.next 

        # Case where we reach end of the list: current = None 
        if current == None:
            print('no data to delete')
            return 
        
        # Case where we found a match current.data == data
        previous_node.next = current.next 
    
    def print_list(self):
        """
        Iterate over nodes in linked list, print the value of each node.
        """
        current = self.head_node
        while current:
            print(current.data)
            current = current.next 


# Create Linked List    
myList = simpleLinkedList()

# Add 1 to the List as first value 
myList.append(1)
print("Head_Node Object: ", myList.head_node)
print("Head_Node Value: " , myList.head_node.data)

# Append numbers 2-4 to linked list 
for i in range(2,5):
    myList.append(i)

# Print values in list
print("")
print("Node object at position 2: ", myList.head_node.next) # prints node in location 2
print(myList.head_node.next.data) # prints data of node in location 2 in list 
print(myList.head_node.next.next.data) # start from node and traverse linked list 


# Prepend values
myList.pre_pend(0)
myList.pre_pend(-1)
myList.pre_pend(-2)


# Print Linked List
print("")
print("Current List:")
myList.print_list()
```
