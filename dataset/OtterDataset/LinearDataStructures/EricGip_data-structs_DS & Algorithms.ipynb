Queues 

Queues are exactly like lines,
First in, First out.

If you enter the line sooner, you exit the line sooner.

First element in queue = head
Last element in queue = tail

Usually, you can only add to the Tail of the queue, and remove the Head element.

Add to tail of queue - Enqueue
Remove from head of queue - Dequeue
When you check the head of the queue but don't move it - Peak

2 Special types of queues: 

Deques - Double ended Queues
    - you can enqueue or dequeue either end, the head or the tail.
    - kind of a hybrid between stacks adnd queues.
    
Priority Queue -
    - you assign each element a priority
    - when you dequeue, you remove elements of highest priority first.
    - if dequeueing two elements with same priority, you will dequeue the oldest element still. 
    
Thankfully, queues are kind of built into python already 

-- from collections import deque
    deque - double ended ques.

```python
"""
Making a queue class from a list
"""

class Queue:
    def __init__(self, head=None):
        self.storage = [head]
    
    def enqueue(self, new_element):
        self.storage.append(new_element)
    
    def peek(self):
        return self.storage[0]
    
    def dequeue(self):
        return self.storage.pop(0)
```

Binary Search

Lets say you have an array sorted in numerical order, you want to check if an integer is in this array.

If you start at the beginning and check every number in the array - O(n) 
    same happends if you start at the other end.
    
Since you have a sorted array, what if you just started at the middle and checked if it was bigger or smaller than the middle number? 
    If its lower, you check to the left
    if its higher, you check to the right.
    
Algorithm = high level description of a trick for solving a problem.

This is called Binary Search.
  when you start the search in the middle and check if its higher or lower.
