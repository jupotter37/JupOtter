# Algorithm

**Algorithm**: an exact prescription specifying the computational process leading from the initial data to the desired result.

Any algorithm is

- Finite

- Deterministic

- Clear and unambiguous

### Example: Euclidean algorithm

**Input**: a, b - integer

**Output**: c - greatest common divisor (GCD) of a and b

1. If $a < b$, then swap them
2. Divide a by b, get the remainder r
3. If $r = 0$, then $c = gcd(a,b) = b$
4. If $r \neq 0$, replace a with b, b with r and return to step 2

## Turing Machine

$a_i$ is a character from the alphabet. $A_p = \{a_1, a_2,..., a_p\}$

$w$ is a word of finite length that consists of the characters $a_i$; $A^*$ is the set of all words.

For any pair of alphabets A and B, it is possible to encode alphabet A using alphabet B and vice versa. Consequently, we can use one alphabet. Usually $𝐴_2 = \{0,1\}$ is considered.

The algorithm is $F: A^* \rightarrow A^*$ 
1. An endless tape divided into cells. Cells that have not been written before are assumed to be filled with the blank symbol $\Lambda$\
2. A head that can read and write symbols on the tape and move the tape left and right one (and only one) cell at a time.
3. $Q = \{q_1, q_2, ..., q_s\}$: alphabet of states
4. Initial state $𝑞_0$, stop state $q_S$
6. A finite table of instructions:

$(q_i, a_j) \rightarrow (q_{i1}, a_{j1}, d)$

where $d = \{L,R\}$

## Big O Notation

The worst case time complexity or the longest amount of time an algorithm can possibly take to complete.

A set of functions 

$Ο(g(n)) = \{f(n) : \exists c > 0, n0: \forall n > n_0, 0 ≤ f(n) ≤ c \cdot g(n)\}$

### Examples:

$f(n) \in O(n)  \rightarrow f(n) = O(n)$

$f(n) = O(1)$

$f(n) = O(n^2)$

$f(n) = O(logn)$

$f(n) = O(e^n)$

$f(n) = O(n \cdot logn)$

![alternatvie text](https://www.tutorialspoint.com/data_structures_algorithms/images/big_oh.jpg)

### $f(n) = O(n^2)$ example:

```python
# Define functions
import time
def function_1(i,j):
    a = i+j
    b = i-j
    c = i*j
    return (a, b, c)

def function_n2(n):
    for i in range(n):
        for j in range(n):
            result = function_1(i,j)
```

### Measure execution time

```python
n_s = [size for size in range(0, 10000, 500)]
time_measures_n2 = []
time_measures_1 = []

for n in n_s:
    start_time = time.time()
    function_n2(n)
    time_measures_n2.append(time.time() - start_time)
    
    start_time = time.time()
    function_1(0, 0)
    time_measures_1.append(time.time() - start_time)    
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(2, figsize=(5, 10))

ax[0].plot(n_s, time_measures_1)
ax[0].set_xlabel('n', fontsize = 14)
ax[0].set_ylabel('time', fontsize = 14)
ax[0].set_title('function_1 execution time', fontsize = 14)

ax[1].plot(n_s, time_measures_n2)
ax[1].set_xlabel('n', fontsize = 14)
ax[1].set_ylabel('time', fontsize = 14)
ax[1].set_title('function_n2 execution time', fontsize = 14)

plt.show()
```

## Average complexity

$X_s = {x_i}$ - set of all possible inputs of size s

Each input needs time $C_{x_{i}}$ 

If we have N input data units of size s and with probability $P_{x_{i}}$, then the average complexity is defined as the mathematical expectation:

$T(s) = \sum_{x_{i} \in X_s}C_{x_{i}}P_{x_{i}}$


## Omega Notation

The lower bound of an algorithm's running time. It measures the best case time complexity, or the best amount of time an algorithm can possibly take to complete.

A set of functions 

$\Omega(g(n)) = \{f(n) : \exists c > 0, n0: \forall n > n_0, 0 ≤  c \cdot g(n)  ≤ f(n)\}$

## Search in array

### Linear search

$T = O(n)$


```python
# Linear search
element = 7
for i in range(len(my_array)):
    if my_array[i] == element:
        print(f"Found at {i} index")
        break
```

### Binary search

Binary search (using ***sorted array***)

$T(n) = O(log(n))$

### Example: 

Search for element "7":

- compare mid element with 7 on each iteration

- move first or last pointer depending on the result

- calculate new mid

![Title](img/binary.png)


```python
# Binary search

my_array_sorted = sorted(my_array)
element = 7
first = 0
last = len(my_array_sorted)
while first < last:
    mid = (first + last) // 2
    if my_array_sorted[mid] < element:
        first = mid + 1
    else:
        last = mid
        
if first == len(my_array_sorted) or my_array_sorted[first]!=element:
    print("Not found")
else:
    print(f"Found at {first} index") 
```

# List

Each element in **singly linked list** contains data and a pointer to the next node. Head points to the first node.

Each node in **doubly linked list** contains data and two pointers: to the next and to the previous elements.

Construct - $O(1)$

Access an element by index: $O(n)$

Find an element: $O(n)$

PushBack: $O(1)$, if we keep a pointer to the last element

PopBack: $O(1)$, if we keep a pointer to the last element

PushFront: $O(1)$

PopFront: $O(1)$

Insert an element to the position specified by pointer:  $O(1)$

![Title](img/list.png)

```python
from collections import deque
llist = deque(['a','b','c'])
llist
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
    def __str__(self):
        return str(self.data)

class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
        self.last = None
        
    def __str__(self):
        node = self.head
        nodes = []
        while node is not None:
            nodes.append(node.data)
            node = node.next
        nodes.append("None")
        return " -> ".join(nodes)
    
    def append(self, element):
        new_node = Node(element)
        if not self.head:
            self.head = new_node
            self.last = new_node
        else:
            self.last.next = new_node
            self.last = new_node

llist = LinkedList()
llist.append('a')
llist.append('b')
llist.append('c')
print(llist)
```

## Binary heap

**A binary heap** is the binary tree type data structure that satisfies the following conditions:

- The value at any node is not less than the value of its children

- The distance from each leaf to the root differs by no more than 1

- The last layer is filled from left to right

A heap can be stored as an array. If $n$ is the number of elements in the heap, then the number of layers is $O(logn)$:

Number of elements in layer 0: 1

Number of elements in layer 1: 2

Number of elements in layer 2: 4

...

Number of elements in layer k: $2^k$

$1+2+4+...+2^k = 2^{k+1}-1 \leq n$

![Title](img/binary_heap.png)

An element can be placed to the correct position in heap using sift up or sift down procedures in $O(logn)$:

### Sift down

![Title](img/sift_down.png)


```python
def SiftDown(heap_array, i):
    left = 2*i + 1
    right = 2*i + 2
    largest = i
    if left < len(heap_array) and heap_array[left] > heap_array[i]:
        largest = left
    if right < len(heap_array) and heap_array[right] > heap_array[largest]:
        largest = right
    if largest!= i:
        heap_array[i], heap_array[largest] =  heap_array[largest], heap_array[i]
        SiftDown(heap_array, largest)

heap_array = [15, 2, 8, 11, 6, 4, 7, 1, 3, 5]
print(SiftDown(heap_array, 1))
heap_array
```


### Sift up

![Title](img/sift_up.png)

```python
def SiftUp(heap_array, i):
    while i > 0:
        parent = (i-1)//2
        if heap_array[i] <= heap_array[parent]:
            return
        heap_array[i], heap_array[parent] = heap_array[parent], heap_array[i]
        i = parent

heap_array = [16, 11, 8, 3, 6, 4, 7, 1, 15, 5]
print(SiftUp(heap_array, 8))
heap_array
```

### Building a heap

Call SiftDown for each element from index [n/2] - 1 to 1 (reversed order) - $O(n)$

At first sight, it takes about $O(nlogn)$ for $n$ elements in the array and $logn$ operation for each SiftDown. However, if we calculate it more carefully, it will be about $O(n)$ because fewer operations are performed on lower nodes.

### Adding an element

Add a new element to the tail of the heap and call SiftUp  - $O(logn)$

### Pop maximum element

1. Save the root element
2. Replace the root element with the tail element
3. Call SiftDown for root

$T(n) = O(logn)$

```python
class BinHeap:
    def SiftDown(self, i):
        left = 2*i + 1
        right = 2*i + 2
        largest = i
        if left < len(self.heap_array) and self.heap_array[left] > self.heap_array[i]:
            largest = left
        if right < len(self.heap_array) and self.heap_array[right] > self.heap_array[largest]:
            largest = right
        if largest!= i:
            self.heap_array[i], self.heap_array[largest] = self.heap_array[largest], self.heap_array[i]
            self.SiftDown(largest)
            
    def SiftUp(self, i):
        while i > 0:
            parent = (i-1)//2
            if self.heap_array[i] <= self.heap_array[parent]:
                return
            self.heap_array[i], self.heap_array[parent] = self.heap_array[parent], self.heap_array[i]
            i = parent
        
    def __init__(self, array):
        self.heap_array = array
        for i in reversed(range(len(self.heap_array)//2)):
            self.SiftDown(i)
            
    def add(self, element):
        self.heap_array.append(element)
        self.SiftUp(len(self.heap_array) - 1)

bin_heap = BinHeap([1, 3, 5, 4, 7, 6, 8, 16, 11, 15])
print(bin_heap.heap_array)
bin_heap.add(20)
print(bin_heap.heap_array)
```
