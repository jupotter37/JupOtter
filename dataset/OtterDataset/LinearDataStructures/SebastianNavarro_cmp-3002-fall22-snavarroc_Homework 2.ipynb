```python
#Clase Node para Linked List
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None):
        self.val = val
        self.next_node = None
    
    def set_next_node(self, next_node):
        self.next_node = next_node
        
class Singly_linked_list:
    """
    Implementation of a singly linked list
    """
    def __init__(self, head_node=None):
        self.head_node = head_node
        
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val)
            node = node.next_node

class Singly_linked_list(Singly_linked_list):  
    #Borra el primer elemento de la lista enlazada
    def delete_head(self): #Complexity O(1)
        # delete to the head
        # A -> B -> null
        # B -> null
        self.head_node = self.head_node.next_node
    
    #Borra el ultimo elemento de la lista enlazada
    def delete_tail(self): #Complexity O(1)
        # delete to the tail
        # A -> B -> null
        # A -> null 
       prev=self.head_node
       node = prev.next_node
       while node.next_node!=None:
             prev=node
             node = node.next_node   
       prev.set_next_node(node.next_node)
    
    #Borra el elemento del medio de la lista enlazada
    def delete_middle(self,value): #Complexity O(1)
        # A -> B -> C -> null
        # A -> null -> C
        prev=self.head_node
        node = prev.next_node
        
        if prev.val == value:
            self.head_node=self.head_node.next_node
            prev.next_node=None
        
        else:
            while node != value:
                prev = node
                node = node.next_node	
                
                if node==None:
                    break
                
            if node:
                prev.set_next_node(node.next_node)
                node.next_node=None
            
            else:
                print("No se encuentra")
                
    @timer
    def reverse(self): #Complexity O(1)
    """
    Reverse the linked list
    """
        sleep(1)
        new_node = self.head_node
        prev = None
        
        while new_node:         
            next_node = new_node.next_node
            new_node.next_node =prev
            prev=new_node
            new_node=next_node
        self.head_node=prev
        
m1 = Node("A")
m2 = Node("B")
m3 = Node("C")
m4 = Node("D")

m1.set_next_node(m2)
m2.set_next_node(m3)
m3.set_next_node(m4)

list1 = Singly_linked_list(m1)
print("List 1:")
list1.list_traversed()
print("\n")
list1.reverse()
print("List 1 reversed:")
list1.list_traversed()
```

## 3. Implement the class Queue using stacks.


```python
#Using stacks for Queue implementation
import ctypes
class Queue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.queue = self._create_queue(self.n) 
        self.stack_1=[]
        self.stack_2=[]
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()
    
class Queue(Queue):
    @timer
    def enqueue(self, item): #Complexity O(1)
        """
        Add new item to the queue
        """
        sleep(1)
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.stack_1.append(item)
        self.l += 1
        print("Element added to the Queue: " + str(item))
    
    @timer
    def dequeue(self): #Complexity O(n)
        """
        Remove an element from the queue
        """
        sleep(1)
        # if both the stacks are empty
        if len(self.stack_1) == 0 and len(self.stack_2) == 0:
            raise ValueError('stack is empty')
 
        # if s2 is empty and s1 has elements
        if len(self.stack_2)==0:
            while self.stack_1:
                self.stack_2.append(self.stack_1.pop())
        self.l -= 1
        return self.stack_2.pop()
    
    @timer
    def first(self): #Complexity O(1)
        """
        Show the first element of the queue
        """
        sleep(1)
        if(len(self.stack_1)==0):
            return self.stack_2[len(self.stack_2)-1]
        else:
            return self.stack_1[0]
    
    @timer
    def full(self): #Complexity O(1)
        """
        Is the queue full?
        """
        sleep(1)
        if self.l == self.n:
            return True
        return False     

    @timer
    def empty(self): #Complexity O(1)
        """
        Is the queue empty?
        """
        sleep(1)
        if self.l == 0:
            return True
        return False
    
    @timer
    def size(self): #Complexity O(1)
        """
        Return size of the queue
        """
        sleep(1)
        return self.l
    
q = Queue(3)
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print("\n")
print('Is Queue empty? ' + str(q.empty())+"\n")
print('Queue has a size of: ' + str(q.size())+"\n")
print('Is Queue full? ' + str(q.full())+"\n")
print('Queue s first item is: ' + str(q.first())+"\n")
print("Element removed from the Queue: " + str(q.dequeue())+"\n")
print('Queue s first item is ' + str(q.first())+"\n")
print("Element removed from the Queue: " + str(q.dequeue())+"\n")
print("Element removed from the Queue: " + str(q.dequeue())+"\n")
print('Is Queue empty? ' + str(q.empty())+"\n")
print('Queue has a size of: ' + str(q.size())+"\n")
print('Is Queue full? ' + str(q.full())+"\n")          
```

## 4. Complete the PriorityQueue class, so that when we call dequeue, the element with the highest priority will be returned. Analyze the complexity of runtime and memory of the enqueue and dequeue methods.

```python
import ctypes
class PriorityQueue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.queue = self._create_queue(self.n)        
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()
    
    @timer
    def enqueue(self, item): #Complexity O(1)
        """
        Add new item to the queue
        """
        sleep(1)
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.queue [self.l] = item
        self.l += 1
    
    @timer
    def dequeue(self): #Complexity O(n^2)
        """
        Return the queue s item with the highest priority 
        """
        sleep(1)
        self.sort()
        return self.queue [0]
    
    #First element represents the priority (while less value, higher priority)
    #Second element represents the value
    #(Priority, value)
    
    @timer
    def insert(self, item): #Complexity O(n^2)
        """
        Add the item to the queue in the correct order
        """
        self.enqueue(item)
        sleep(1)
        l = self.l
        for i in range(1, l-1):
            if(item [0] >= self.queue [i-1] [0] and item [0] <= self.queue [i] [0]): 
                for j in range (l-1, i,-1):
                    self.queue [j] = self.queue [j-1]   
                self.queue [i] = item
                break
    
    @timer
    def sort(self): #Complexity O(n^2)
        """
        Sort the queue using the algorith of insertion sort
        """
        sleep(1)
        for j in range (0, self.l):
            key = self.queue [j]
            i = j - 1
            while i >= 0 and self.queue [i] > key:
                self.queue [i + 1] = self.queue [i]
                i = i - 1
            self.queue [i + 1] = key  

q = PriorityQueue(10)
q.enqueue((9,2))
q.enqueue((1,5))
q.enqueue((8,4))

print("Queue sorted:")
q.sort()
print(q.queue[0:q.l])
print("\nQueue inserted (2,3):")
q.insert((2,3))
print(q.queue[0:q.l])
print("\nQueue s element with highest priority:")
print(q.dequeue())
```

En efecto, se observa que la operacion de $enqueue$ toma mucho menos tiempo que $dequeue$, ya que $dequeue$ utiliza el algoritmo de $insertion$ $sort$ para ordenar cuya complejidad es $O(n^2)$, y luego retorna el primer elemento de la cola que corresponde al de mayor prioridad, por tanto su complejidad final es $O(n^2)$, a diferencia de la operacion de $enqueue$ cuya complejidad es de apenas $O(1)$. Se tiene una diferencia de aproximadamente $1.0049 [seconds]$. Por otra parte, la operacion $insertion$ tiene una complejidad $O(n^2)$ al igual que el algoritmo de insertion sort, ya que utiliza dos lazos anidados, sin embargo realiza una mayor cantidad de operaciones al mover elementos del $queue$. Por ende, la diferencia entre la operacion $sort$ e $insertion$ es de $1.0312 [seconds]$

### 5. A given linked-list (singly or doubly) represents an integer number. For example, 345 is represented by the singly-linked list 3 -> 4 ->5. 

```python
#Uso de librerias de Python
import numpy as np
import pandas as pd
from time import sleep
import matplotlib.pyplot as plt
import random
from time import time
import ctypes
from functools import wraps
import time

#Funcion para medir el tiempo de ejecucion
def timer(func):
    @wraps(func)
    def timeit_wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()-1
        total_time = end_time - start_time
        print(f'Function {func.__name__} took: {total_time:.4f} [seconds]')
        return result
    return timeit_wrapper

#Implementacion de la clase nodo para uso de Linked List
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None,next_node=None):
        self.val = val
        self.next_node=next_node
    
    def set_next_node(self, next_node):
        self.next_node = next_node
        
class Singly_linked_list:
    """
    Implementation of a singly linked list
    """
    def __init__(self, head_node=None):
        self.head_node = head_node
            
    def list_traversed(self):
        node = self.head_node
        while node:
            print(node.val, end =' ')
            node = node.next_node
            
    def insert(self, new_node): #Complexity O(1)
        """
        Insert a node into the Linked List
        """
        current = self.head_node
        if current:
            while current.next_node:
                current = current.next_node
            current.next_node = new_node
        else:
            self.head_node = new_node
            
    def reverse(self): #Complexity O(1)
        """
        Reverse a Linked List
        """
        new_node = self.head_node
        prev = None
        
        while new_node:         
            next_node = new_node.next_node
            new_node.next_node =prev
            prev=new_node
            new_node=next_node
        self.head_node=prev
        
@timer  
def transform(num): #Complexity O(1)
    """
    Transform a number to a Linked List
    """
    sleep(1)
    list0=Singly_linked_list()
    for x in str(num):
        list0.insert(Node(int(x)))
    return list0

@timer
def sum_list(list1,list2,list3): #Complexity O(n) ya que depende del numero de digitos de las Linked List
    """
    Returns the sum of 3 Linked List in the Linked List format
    """
    sleep(1)
    list1.reverse()
    list2.reverse()
    list3.reverse()

    node_1=list1.head_node
    node_2=list2.head_node
    node_3=list3.head_node
    st=''
    res=0
    
    while node_1:
       
        sum_1=sum([node_1.val,node_2.val,node_3.val])
        
        if(sum_1+res >= 10) and not (node_1.next_node==None):
            st=st+str(sum_1+res)[1]
            res=int(str(sum_1+res)[0])
            
        elif(sum_1+res < 10) and not (node_1.next_node==None):
            st=st+str(str(sum_1+res))
            res=0
            
        elif(node_1.next_node==None):
            st=st+str(str(sum_1+res))[::-1]
            
        node_1=node_1.next_node
        node_2=node_2.next_node
        node_3=node_3.next_node
        
    ans=transform(st)
    ans.reverse()
    return ans

def generate_number(num): #Complexity O(1)
    """
    Generate a random number with the number of digits specified
    """
    return random.randint(10**(num-1), 10**(num))


print("List 1:")
list1=transform(345)
list1.list_traversed()
print("\nList 2:")
list2=transform(123)
list2.list_traversed()
print("\nList 3:")
list3=transform(567)
list3.list_traversed()
print("\nList 1 + List 2 + List 3:")
list4=sum_list(list1,list2,list3)
list4.list_traversed()

print("\n\nFor numbers with 70 digits:")
print("List 1:")
list5=transform(generate_number(70))
list5.list_traversed()
print("\nList 2:")
list6=transform(generate_number(70))
list6.list_traversed()
print("\nList 3:")
list7=transform(generate_number(70))
list7.list_traversed()
print("\nList 1 + List 2 + List 3:")
list8=sum_list(list5,list6,list7)
list8.list_traversed()
```
