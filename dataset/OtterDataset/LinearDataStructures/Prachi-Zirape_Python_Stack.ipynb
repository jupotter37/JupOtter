```python
#Date-10/11/22
#stack using Array
from sys import maxsize
class StackUsingArray:
    def __init__(self):
        self.stack=[]
        
#push operation  add an item to stack. It increases size by 1    
    def pushEl(self,element):
        self.stack.append(element)
        
#pop operation        
    def pop(self):
        if not(self.isEmpty()):
            lastElelement=self.stack[-1]
            del(self.stack[-1])
            return lastElelement
        else:
            return -1
        
#check if its empty empty when stack size is 0
    def isEmpty(self):
        return self.stack==[]
    
    def prinStack(self):
        print(self.stack)
        
sa=StackUsingArray()
sa.pop()
sa.prinStack()
sa.pushEl(10)
sa.pushEl(20)
sa.pushEl(30)
sa.pushEl(40)
sa.prinStack()
sa.pop()
sa.prinStack()
```

```python
#1-Queue using Stacks
#making Enqueue OPeration COstly 

class Queue:
    # two stacks with costly enQueue()
    def __init__(self):
        self.s1=[]
        self.s2=[]
        
    def enQueue(self,x): #its cause Queues insertion happens at rear end and deletion from front
        # Move all elements from s1 to s2
        while(len(self.s1))!=0:
            self.s2.append(self.s1[-1])
            self.s1.pop()
            
        # Push item into self.s1
        self.s1.append(x)
        
        # Push everything back to s1
        while(len(self.s2)!=0):
            self.s1.append(self.s2[-1])
            self.s2.pop()
        
    # Dequeue an item from the queue
    def deQueue(self):
        # if first stack is empty
        if len(self.s1)==0:
            return -1
        x=self.s1[-1]
        self.s1.pop()
        return x
if __name__ == '__main__':
    q = Queue()
    q.enQueue(1)  #rear
    q.enQueue(2)
    q.enQueue(3)  #front
 
    print(q.deQueue())
    print(q.deQueue())
    print(q.deQueue())
```

```python
#making Dequeue OPeration COstly 

class Queue:
    def __init__(self):
        self.s1=[]
        self.s2=[]  #two stack arrays
        
    #EnQueue item to the queue
    def enQueue(self,x):
        self.s1.append(x)
    
    # DeQueue item from the queue
    def deQueue(self):
        # if both the stacks are empty
        if len(self.s1)==0 and len(self.s2)==0:
            return -1
        
        # if s2 is empty and s1 has elements
        elif len(self.s2)==0 and len(self.s1)>0:
            while(len(self.s1)):
                temp=self.s1.pop()
                self.s2.append(temp)
                #print("K",len(self.s2))
            return self.s2.pop()  #popping right aftr enqueue complete
        else:
            #print("k",len(self.s2))
            return self.s2.pop()  #popping after enque operation complte
        
if __name__ == '__main__':
    q = Queue()
    q.enQueue(1)
    q.enQueue(2)
    q.enQueue(3)
 
    print(q.deQueue())
    print(q.deQueue())
    print(q.deQueue())
```

```python
#recursive
class Queue:
    def __init__(self):
        self.s=[]
        
     #Enqueue an item to the queue
    def enQueue(self,data):
        self.s.append(data)
    
    #Dequeue items from queue
    def deQueue(self):
        #retrun if queue is empty
        if len(self.s)<=0:
            return -1
        # pop an item from the stack
        x=self.s[len(self.s)-1]
        self.s.pop()
       
        # if stack become empty
        # return the popped item
        if len(self.s)<=0:
            return x      
         # recursive call
        item=self.deQueue()  #Saving items in recursive memory calling function over and ovr until q gets empty
        
        # push popped item back to
        # the stack
        self.s.append(x)
              
        return item
    
if __name__ == '__main__':
    q = Queue()
    q.enQueue(1)
    q.enQueue(2)
    q.enQueue(3)
     
    print(q.deQueue())
    print(q.deQueue())
    print(q.deQueue())
```
