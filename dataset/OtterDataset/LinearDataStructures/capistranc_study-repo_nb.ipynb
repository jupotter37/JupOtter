````julia
class ListNode {
    int val;
    ListNode next;
    ListNode(int x) {
        val = x;
        next = null;
    }
}
var n = new ListNode(1);

public class Solution {
    public ListNode detectCycle(ListNode head) {
        ListNode sp=head, fp=head;
        while (fp != null) {
            sp = sp.next;
            fp = fp.next == null ? null : fp.next.next;
            if (sp == fp) break;
        }
        if (fp == null) return null;
        fp = head;
        while (fp != sp) {
            sp = sp.next;
            fp = fp.next;
        }
        return sp;
    }
}

new Solution().detectCycle(n)
````

````julia
/*
Stack of Plates: 
Imagine a (literal) stack of plates. If  the stack gets too high, it might topple. Therefore, in 
real life, we would likely start a new stack when the previous stack exceeds some threshold. 
Implement a data structure SetOfStacks that mimics this. SetO-fStacks should be composed of several 
stacks and should create a new stack once the previous one exceeds capacity. 
SetOfStacks. push() and SetOfStacks. pop() should behave identically to a single stack 
(that is, pop () should return the same values as it would if there were just a single stack). 

FOLLOW UP: Implement a function popAt ( int  index) which performs a pop operation on a specific 
sub-stack. 

SOLN:
Just have a vec of stacks.

Pop removes from top of rightmost stack.  When that stack is empty pop it.  Push adds to rightmost 
stack.  When that stack is at capacity, add a new stack to right of it.

Examples:
cap = 3, add [1,2,3,4,5], pop()x3, add [6,7,8,9,0], pop()*5

[1,2,3],[4,5]
[1,2]           5,4,3
[1,2,6],[7,8,9],[0]
[1,2]           0,9,8,7,6

Complexity:
Time: O_pop(1),O_push(1)
Space: O(N)
*/
````

````julia
#[derive(Debug)]
struct DinnerPlates {
    capacity: i32,
    stacks: Vec<Vec<i32>>
}

/** 
 * `&self` means the method takes an immutable reference.
 * If you need a mutable reference, change it to `&mut self` instead.
 */
impl DinnerPlates {

    fn new(capacity: i32) -> Self {
        Self{capacity: capacity,stacks: vec![]}    
    }
    
    fn push_ctci(&mut self, val: i32) {
        match self.stacks.last() {
            None => self.stacks.push(vec![val]),
            Some(v) => {
                // let u: () = v;
                if v.len() >= self.capacity as usize {
                    self.stacks.push(vec![val]);
                } else {
                    v.push(val);
                }
            },
        } 
    }    
    
    fn pop(&mut self) -> i32 {
        return match self.stacks.last() {
            None => -1,
            Some(v) => match v.pop() {
                None => {
                    self.stacks.pop();
                    DinnerPlates::pop(self)
                },
                Some(val) => val,
            }
        } 
    }
    
    fn pop_at_stack(&mut self, index: i32) -> i32 {
        return match self.stacks.get(index as usize) {
            None => -1,
            Some(v) => match v.pop() {
                None => -1,
                Some(val) => val
            }
        }
    }
}
````

````julia
class Solution {
    public boolean isSubtree(TreeNode T1, TreeNode T2) {
        class Helpers {
            boolean eqTree(TreeNode T1, TreeNode T2) {
                if (T1 == null) return T2 == null;
                if (T2 == null) return T1 == null;
                return T1.val == T2.val && eqTree(T1.left,T2.left) && eqTree(T1.right,T2.right);
            }
        }
        if (T1 == null) return T2 == null;
        return new Helpers().eqTree(T1,T2) || isSubtree(T1.left,T2) || isSubtree(T1.right,T2);
    }
}
var T1 = new TreeNode().makeTree(new Integer[]{4,3,7,6,9,null,2,0,8,null,3,null,null,null,1});
var T2 = new TreeNode().makeTree(new Integer[]{3,6,9,0,8,null,3});
System.out.println(new Solution().isSubtree(T1,T2));
var T1 = new TreeNode().makeTree(new Integer[]{4,3,7,6,9,null,2,0,8,null,3,null,null,null,1});
var T2 = new TreeNode().makeTree(new Integer[]{3,6,9,1,8,null,3});
System.out.println(new Solution().isSubtree(T1,T2));
var T2 = new TreeNode().makeTree(new Integer[]{3,6,9,0,8,null,3,1});
var T1 = new TreeNode().makeTree(new Integer[]{4,3,7,6,9,null,2,0,8,null,3,null,null,null,1});
System.out.println(new Solution().isSubtree(T1,T2));
var T1 = new TreeNode().makeTree(new Integer[]{4,3,7,6,9,null,2,0,8,null,3,null,null,null,1});
var T2 = new TreeNode().makeTree(new Integer[]{3,6,9,0,8,1,3});
System.out.println(new Solution().isSubtree(T1,T2));

var T1 = new TreeNode().makeTree(new Integer[]{3,4,5,1,null,2});
var T2 = new TreeNode().makeTree(new Integer[]{3,1,2});
System.out.println(new Solution().isSubtree(T1,T2));
/*
                3
            /       \
        4               5
    /               /
1                 2

                    3
                /       \
            1               2
*/
````

````julia
/*
First Common Ancestor: 
Design an algorithm and write code to find the first common ancestor of two nodes in a binary tree. Avoid storing 
additional nodes in a data structure. NOTE: This is not necessarily a binary search tree. 
Hints: #70, #76, #28, #36, #46, #70, #80, #96

Soln:
Recursively traverse tree - return whether subtree has `n1`, subtree has `n2`, 
and current LCA, as well as current LCA depth.  
Note: if node.val was not unique we would also have to track
LCA depth b/c LCA in different sides of tree may have differnet depths

current subtree contains n1 = n.val == n1 || of left and right
... same for n2
LCA = LCA_right.depth > LCA_left.depth ? LCA_right : LCA_left

If subtree contains n1 and n2
- if LCA null, return current node
- otherwise return current LCA
else:
    return null


Examples:
[3,4,1,1,7, ,5,0,5, ,8, ,4]

                    3
                  /   \
                4       1
              /   \       \
            1       7       5
          /   \       \       \
        0       5       8      4

n = 1, n = 5

Complexity:
Time: O(N)
Space: O(depth)
*/
````

## [8.1 Triple Step](https://leetcode.com/problems/n-th-tribonacci-number/)

````julia
/*
Triple Step: 
A child is running up a staircase with n steps and can hop either 1 step, 2 steps, or 3 steps at a time. 
Implement a method to count how many possible ways the child can run up the stairs. 
Hints: #152, #178, #217, #237, #262, #359

SOLN:
can solve w/ recursion.  
f(i) = # of ways to get to stair i
f(i) = f(i-1) + f(i-2) + f(i-3)
f(0) = 1
f(i < 0) = 0
f(1) = f(0) + f(-1) + f(-2) = f(0) = 1
f(2) = f(1) + f(0) + f(-1) = f(1) + f(0) = 2
.....

Complexity:
Time = O(N)
Space = O(N)
*/
````

````julia
class Solution {
    long nWays(int N) {
        long[] dp = new long[Math.max(3,N+1)];
        dp[0] = 0; dp[1] = 1; dp[2] = 1;
        for (int i = 3; i < N+1; i++) {
            dp[i] = dp[i-1] + dp[i-2] + dp[i-3];
        }
        return dp[N];
    }

    public int tribonacci(int n) {
        return (int)nWays(n);   
    }
}
for (int i = 0; i < 10; i++) {
    System.out.println(new Solution().nWays(i));
}
````

## [8.2 Robot in a Grid](https://leetcode.com/problems/unique-paths-ii/)


````julia
/*
Robot in a Grid: Imagine a robot sitting on the upper left corner of grid with r rows and c columns. The robot can only move in two directions, right and down, but certain cells are "off limits" such that the robot cannot step on them. Design an algorithm to find a path for the robot from the top left to the bottom right. Hints: #331, #360, #388 

SOLN:
solve recursively:
npaths(r,c) = {
    if passable: npaths(r-1) + npaths(c-1)
    else (impassable or OOB): 0
}

Complexity:
Time:   O(H*W)
Space:  O(H*W) or O(1) in place
*/
````

````julia
class Solution {
    int npaths(int[][] obstacles) {
        int H = obstacles.length, W = obstacles[0].length;
        var dp = new int[H][W];
        dp[0][0] = obstacles[0][0] == 1 ? 0 : 1;
        for (int r = 1; r < H; r++) dp[r][0] = obstacles[r][0] == 1 ? 0 : dp[r-1][0];
        for (int c = 1; c < W; c++) dp[0][c] = obstacles[0][c] == 1 ? 0 : dp[0][c-1];
        for (int r = 1; r < H; r++)
            for (int c = 1; c < W; c++)
                dp[r][c] = obstacles[r][c] == 1 ? 0 : dp[r-1][c] + dp[r][c-1];
        return dp[H-1][W-1];
    }
    public int uniquePathsWithObstacles(int[][] obstacleGrid) {
        return npaths(obstacleGrid);   
    }
}
````

## [8.4 Power Set](https://leetcode.com/problems/subsets/):

````julia
/*
Write a method to return all subsets of a set.

SOLN:
Recursively construct powerset - for each element in set we take the powerset of the
set suffix.  Then for each element of suffix powerset, we duplicate and add current
set el to the copy of that set.  This is then the powerset of of the set suffix and
the current el in the set.

For empty set, powerset has a single element, which is the empty set.

Examples:


Complexity:
Time:   O(N*2^N)
Space:  O(N*2^N) 
*/
````

````julia
import java.util.stream.Collectors;

class Solution {
    ArrayList<ArrayList<Integer>> powerset(List<Integer> S) {
        var ret = new ArrayList<ArrayList<Integer>>();
        if (S.size() == 0) ret.add(new ArrayList<Integer>());
        else {
            var i = S.remove(S.size()-1);
            var prefix_powerset = powerset(S);
            for (ArrayList<Integer> s: prefix_powerset) {
                ret.add(new ArrayList<Integer>(s));
                s.add(i);
                ret.add(s);
            }
        }
        return ret;
    }

    ArrayList<ArrayList<Integer>> powersetSorted(List<Integer> S) {
        var ret = new ArrayList<ArrayList<Integer>>();
        if (S.size() == 0) ret.add(new ArrayList<Integer>());
        else {
            var n = S.remove(S.size()-1);
            var prefix_powerset = powersetSorted(S);
            var inclusive_prefix_powerset = prefix_powerset;
            var exclusive_prefix_powerset = new ArrayList<ArrayList<Integer>>();
            for (ArrayList<Integer> s: prefix_powerset) {
                exclusive_prefix_powerset.add(new ArrayList<Integer>(s));
                s.add(n);
            }
            class Locals {
                int cmp(ArrayList<Integer> l1, ArrayList<Integer> l2) {
                    if (l1.size() == 0) return -1;
                    if (l2.size() == 0) return 1;
                    for (int i = 0; i < l1.size() && i < l2.size(); i++) {
                        if      (l1.get(i) < l2.get(i)) return -1;
                        else if (l1.get(i) > l2.get(i)) return 1;
                    }
                    return l1.size() - l2.size();
                }
            }
            int i = 0, j = 0;
            for (; i < inclusive_prefix_powerset.size() && j < exclusive_prefix_powerset.size();) {
                var s = inclusive_prefix_powerset.get(i);
                var t = exclusive_prefix_powerset.get(j);
                if (new Locals().cmp(s,t) < 0) {
                    ret.add(s);
                    i++;
                } else {
                    ret.add(t);
                    j++;
                }
            }
            ret.addAll(inclusive_prefix_powerset.subList(i,inclusive_prefix_powerset.size()));
            ret.addAll(exclusive_prefix_powerset.subList(j,exclusive_prefix_powerset.size()));
            System.out.println(inclusive_prefix_powerset);
            System.out.println(exclusive_prefix_powerset);
            System.out.println(ret);
        }
        return ret;
    }
    

    public List<List<Integer>> subsets(int[] nums) {
        List l = new ArrayList<Integer>();
        for (int n : nums) l.add(new Integer(n));
        // return powerset(l);
        return powersetSorted(l);   
    }
}

var l = new ArrayList<Integer>();
l.addAll(Arrays.asList(new Integer[]{1,2,3,4,5}));
System.out.println(l);
// new Solution().powerset(l)
new Solution().subsets(new int[]{1,2,3,4,5})
// new Solution().subsets(new int[]{1,2,3})
````

## [8.13 Stack of Boxes:](https://leetcode.com/problems/maximum-height-by-stacking-cuboids/)

````julia
/*
Given n cuboids where the dimensions of the ith cuboid is cuboids[i] = [widthi, lengthi, heighti] (0-indexed). 
Choose a subset of cuboids and place them on each other.

You can place cuboid i on cuboid j if widthi <= widthj and lengthi <= lengthj and heighti <= heightj. 
You can rearrange any cuboid's dimensions by rotating it to put it on another cuboid.

Return the maximum height of the stacked cuboids.

Solution:
Each cubeoid [h,w,l] can be rotated 3 ways [h,w,l],[w,l,h],[l,h,w] but only used once

What if we didnt order the boxes first?
    [1,1],[4,2],[2,2]
    - then it would be impossible to stack it 1,2,0 like you want to

we could order the boxes by their smallest dimension descending, breaking ties by next smallest and so on,
b/c that guarantees that we would never want to stack an earlier box on a later one 

 If we stack any box i, all boxes prev
may not be stacked (unless their smallest dim is equal) or, more accurately,
have already been considerdd.  So we only need to consider boxes
after box i.

This stacking problem can be solved recursively w/ params box index i, and dimension limits h,w,l
 -> stack(i,h,w,l)
We then pick the box j after and including i that is smaller than all dimension limits and has greatest
NOTE: try all 3 orientations for that box

MAX (
    box[j].h + stack(j+1,min(h,box[j].h),...),
    box[j].w + stack(j+1,min(h,box[j].w),...)
    box[j].l + stack(j+1,min(h,box[j].l),...)
)
stack(i=boxes.length,:,:,:) = 0

SOLUTION 2:
Problem bounds can be parmaeteried by h,w,l , but they may also be parameterized by previous box j we 
stacked.  

So for box i, the maximum stack height is
MAX(over j < i,
    {
        if boxes[i] >= boxes[j]:    boxes[i].h + stack(j)
        else                        stack(j)
    })
We dont need to consier j < i b/c those boxes have a dimension that is smaller 
than the smalleset dimension of i

Note also that b/c all dimensions must be smaller than prev box there is no point considering different 
box orientations.  We want max hiehgt so largest box dimemsion will always be used for vertical

If probelm only constrainted stacking by WxL then dp would have an extra dim of len 6 (or 3) where
different orientations of boxes are considered.



Examples:
[[1,1,2],[4,2,3],[2,2,1]] => 7

[3,2,4],[2,1,2],[1,1,2] => 8
[4,3,2],[2,1,2],[2,1,1] => 8

dp[0,4,4,4] = 4 + dp[1,3,2,4]

[1,1,2]
3 * 2 * 1

[h,w,l] == [h,]     -> [h,w,l],[h,l,w]
            [w]     -> [w,h,l],[w,l,h]
            [l,]    -> [l,h,w],[l,w,h]

Complexity:
NOTE: if H,W,L >>> N, then we can use a hash map instead of an array to reduce their domain size to N
Time:   O(N^2*H*W*L)
Space:  O(N^2*H*W*L)
*/
````

````julia
class Solution {
    private int _min(int[] b) {
        int m = Integer.MAX_VALUE;
        for (int n: b) m = Math.min(m,n);
        return m;
    }
    private int _max(int[] b) {
        int m = Integer.MIN_VALUE;
        for (int n: b) m = Math.max(m,n);
        return m;
    }

    int stack(int[][] boxes) {
        for (int[] b: boxes) Arrays.sort(b);
        Arrays.sort(boxes,(int[] b1, int[] b2) -> {
            if      (b1[0] != b2[0]) return b2[0] - b1[0];
            else if (b1[1] != b2[1]) return b2[1] - b1[1];
            else                     return b2[2] - b1[2];
        });
        int D = Integer.MIN_VALUE;
        for(int[] b: boxes) D = Math.max(D,this._max(b));
        int[][][][] dp = new int[boxes.length + 1][D+1][D+1][D+1];
        for (int h = 0; h <= D; h++) {
            for (int w = 0; w <= D; w++) {
                for (int l = 0; l <= D; l++) {
                    for (int i = boxes.length-1; i >= 0; i--) {
                        int ri = Integer.MIN_VALUE;
                        for (int j = i; j < boxes.length; j++) {
                            int hj=boxes[j][0],wj=boxes[j][1],lj=boxes[j][2];
                            int r1=0,r2=0,r3=0,r4=0,r5=0,r6=0;
                            if (hj <= h && wj <= w && lj <= l)  
                                r1 = hj + dp[j+1][Math.min(hj,h)][Math.min(wj,w)][Math.min(lj,l)];
                            if (hj <= h && lj <= w && wj <= l)  
                                r2 = hj + dp[j+1][Math.min(hj,h)][Math.min(lj,w)][Math.min(wj,l)];
                            if (wj <= h && lj <= w && hj <= l)  
                                r3 = wj + dp[j+1][Math.min(wj,h)][Math.min(lj,w)][Math.min(hj,l)];
                            if (wj <= h && hj <= w && lj <= l)  
                                r4 = wj + dp[j+1][Math.min(wj,h)][Math.min(hj,w)][Math.min(lj,l)];
                            if (lj <= h && hj <= w && wj <= l)  
                                r5 = lj + dp[j+1][Math.min(lj,h)][Math.min(hj,w)][Math.min(wj,l)];
                            if (lj <= h && wj <= w && hj <= l)  
                                r6 = lj + dp[j+1][Math.min(lj,h)][Math.min(wj,w)][Math.min(hj,l)];
                            
                            ri = this._max(new int[]{ri,r1,r2,r3,r4,r5,r6});
                            // System.out.println(String.format(
                            //     "i=%d,j=%d,h=%d,w=%d,l=%d,ri=%d,r1=%d,r2=%d,r3=%d,r4=%d,r5=%d,r6=%d",
                            //     i,j,h,w,l,ri,r1,r2,r3,r4,r5,r6));
                        }
                        // System.out.println(String.format("i=%d,h=%d,w=%d,l=%d,ri=%d",i,h,w,l,ri));
                        dp[i][h][w][l] = ri;
                    }
                }
            }
        }
        return dp[0][D][D][D];
    }

    public int maxHeight(int[][] cuboids) {
        return this.stack(cuboids);
    }
}
int[][] boxes = new int[][]{{1,1,2},{4,2,3},{2,2,1}};
System.out.println(new Solution().maxHeight(boxes));
int[][] boxes = new int[][]{{88,45,69},{88,15,96},{49,78,68},{96,72,80},{80,86,13},
                            {9,99,73},{94,42,39},{30,19,58},{10,32,21},{20,52,85}};
System.out.println(new Solution().maxHeight(boxes));
````

````julia
class Solution {
    int stack(int[][] boxes) {
        for (int[] b: boxes) Arrays.sort(b);
        Arrays.sort(boxes,(int[] b1, int[] b2) -> {
            if      (b1[0] != b2[0]) return b2[0] - b1[0];
            else if (b1[1] != b2[1]) return b2[1] - b1[1];
            else                     return b2[2] - b1[2];
        });
        int[] dp = new int[boxes.length];
        for (int i = 0; i < boxes.length; i++) {
            dp[i] = boxes[i][2];
            for (int j = 0; j < i; j++) {
                if ((boxes[i][0] <= boxes[j][0]) && (boxes[i][1] <= boxes[j][1]) && (boxes[i][2] <= boxes[j][2])) {
                    dp[i] = Math.max(dp[i],boxes[i][2] + dp[j]);
                }
            }
        }
        int r = Integer.MIN_VALUE;
        for (int h : dp) r = Math.max(h,r);
        return r;
    }

    public int maxHeight(int[][] cuboids) {
        return this.stack(cuboids);
    }
}
int[][] boxes = new int[][]{{1,1,2},{4,2,3},{2,2,1}};
System.out.println(new Solution().maxHeight(boxes));
int[][] boxes = new int[][]{{50,45,20},{95,37,53},{45,23,12}};
System.out.println(new Solution().maxHeight(boxes));
int[][] boxes = new int[][]{{38,25,45},{76,35,3}};
System.out.println(new Solution().maxHeight(boxes));
int[][] boxes = new int[][]{{7,11,17},{7,17,11},{11,7,17},{11,17,7},{17,7,11},{17,11,7}};
System.out.println(new Solution().maxHeight(boxes));
int[][] boxes = new int[][]{{88,45,69},{88,15,96},{49,78,68},{96,72,80},{80,86,13},
                            {9,99,73},{94,42,39},{30,19,58},{10,32,21},{20,52,85}};
System.out.println(new Solution().maxHeight(boxes));
````
