```python
import pandas as pd
```

```python
df = pd.read_csv("LeetCode_questions.csv")
link = "https://leetcode.com/problems/"
leetcode_links = []
for i in df["Questions"]:
    k = ("-").join(i.split()[1:]).lower()
    print(k)
    leet_link = link + k
    print(leet_link)
    leetcode_links.append(leet_link)
```

```python
df["leetcode_links"] = leetcode_links
df.to_csv("LeetCode_questions_updated.csv", index=False)
```
