## Big O Complexities

```python
import matplotlib.pyplot as plt
import numpy as np

# Input size (e.g., number of elements)
n = np.arange(1, 10)

# Define various Big O time complexities
constant = np.ones_like(n)
logarithmic = np.log(n)
linear = n
linearithmic = n * np.log(n)
quadratic = n**2
cubic = n**3
exponential = 2**n

# Create the plot
plt.figure(figsize=(5, 3))
plt.plot(n, constant, label='O(1) - Constant')
plt.plot(n, logarithmic, label='O(log n) - Logarithmic')
plt.plot(n, linear, label='O(n) - Linear')
plt.plot(n, linearithmic, label='O(n log n) - Linearithmic')
plt.plot(n, quadratic, label='O(n^2) - Quadratic')
plt.plot(n, cubic, label='O(n^3) - Cubic')
plt.plot(n, exponential, label='O(2^n) - Exponential')

# Set plot labels and legend
plt.ylim(0, 2**3) 
plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.yscale('linear') 
plt.title('Big O Notation - Time Complexities')
plt.legend()

# Display the plot
plt.show()
```

### Classes

```python
'''
1. Declare Classes
2. Declare Constructor
3. Create Member Variables
4. Create Methods
5. Declare and Initialize Object Instances
'''

class TreeNode:
  def __init__(self, val, left = None, right = None):
    self.val = val
    self.left = left
    self.right = right

class Tree:
  def __init__(self, root):
    self.root = root

  def dfs(self, root, helper_function):
    if not root: 
      return
      
    helper_function(root.val)
    
    self.dfs(root.left, helper_function)
    return self.dfs(root.right, helper_function)

  def print_nodes_on_runtime(self):
    self.dfs(self.root, lambda val: print(val))
  
  def __str__(self): # custom dunder method for print() & str()
    res = []
    self.dfs(self.root, lambda val: res.append(val))
    return ''.join(map(str, res)) # convert every element in res to string and combine them into one string
    
root = TreeNode('a', TreeNode('b'), TreeNode('c'))
tree = Tree(root)

# Print Tree Nodes (On run-time)
print('Print Tree Nodes (On run-time) :\n')
tree.print_nodes_on_runtime()

# Print Tree
print('Print Tree: ', tree)

# Convert Tree to String
print('Convert Tree to String: ', str(tree))
```

## Arrays
An array is a data structure that stores items in a sequential structure. Arrays have 3 properties: a set size, elements of the same type, and elements stored contiguously in memory. Their sizes have to be predetermined because the computer needs to find a contiguous set of empty memory slots to store the elements side by side. This sequential memory allows any element to be accessed through indexing in O(1) time.

 A dynamic array is an abstract data structure that allows the size to change in run time as opposed to compile time. This is acheived by triggering the computer to find a larger set of contiguous memory slots when the current set is filled up automatically. This can be expensive cause all the data from the previous memory slot has to be copied over to the new memory slot, taking O(n) time.
 
 A list is a high level abstract data structure with dynamic sizing and non-homegnous data types. The only thing they all have in common is that the elements are indexed adjacently in memory.

**Matrix Iteration**

An array can represent a matrix in two ways:

1. Row Major
    1. Each list represents a row
    2. Each element inside of the list represents a column
    3. Positional Order: Matrix[row][col]
2. Column Major
    1. Each list represents a column
    2. Each element inside of the list represents a row
    3. Positional Order: Matrix[col][row]

Iterating over a matrix requires two nested for loops. The first loop will iterate over all the rows. The second loop with iterate over all the columns in each row. This would print every column in one row before moving to the next. The inverse can also be implemented.

What the outer and inner loops are iterating over (row/column) determines whether the matrix is being iterated column-first or row-first. Essentially, the outer loop acts as a ‘pin’, while the inner loop ‘rolls’. The outer loop ‘pins’ the value until the inner loop is done ‘rolling’ through them, and then it ‘pins’ the next value and so on…

1. Row-first Iteration
   1. Loop over row
   2. Loop over each column in row
2. Col-first Iteration
   1. Loop over column
   2. Loop over each row in column

Regardless of whether you're iterating column-first or row-first, the matrix should always be accessed in the order it was represented. Otherwise, it will go out of bounds. 
1. Row major: matrix[r][c]
2. Column major: matrix[c][r]

```python
# Matrix Iteration (Forwards)
mat = [
  [1,2,3,10],
  [4,5,6,20],
  [7,8,9,30]
  ]
  
row = len(mat); # define row boundaries
col = len(mat[0]); # define col boundaries

# Matrix Row First Iteration
row_first = ''
for r in range(row):
   for c in range(col):
      row_first += str(mat[r][c]) # follow the order of the matrix representation
print('Matrix Row First Iteration: ', row_first)

# Matrix Column First Iteration
col_first = ''
for c in range(col):
   for r in range(row):
      col_first += str(mat[r][c]) # follow the order of the matrix representation
print('Matrix Column First Iteration: ',col_first)
```

### Transpose Matrix

```python
def transpose(matrix: list[list[int]]) -> list[list[int]]:
        row = len(matrix)
        col = len(matrix[0])
        transposed_matrix = [[0 for i in range(row)] for j in range(col)]

        for r in range(row):
            for c in range(col):
                transposed_matrix[c][r] = matrix[r][c]

        return transposed_matrix
```

### Rotate Image

```python
def rotate(matrix: list[list[int]]) -> None:
    entries = len(matrix)
    values = len(matrix[0])

    for e in range(entries):
        for v in range(e+1,values):
            matrix[e][v],matrix[v][e] = matrix[v][e],matrix[e][v]
        matrix[e].reverse()

    return matrix
```

### Binary Search

Binary search is a searching algorithm that can efficiently find values in a list given the list is sorted. It does so by taking the midpoint of the list and then deciding if the target value is before or after it. This will inform the algorithm which half to search. It then finds the midpoint of the the half it searches, and so on. This means the search space is divided by half at every iteration. This gives it a time complexity of O(log(n)).

### Sorting

Sorting algorithms organize data in a specified order. The default sorting order considered here will be ascending order. These algorithms are analyzed by the following: number of swaps and comparisons, amount of memory, recursion (or is it iterative?), stability (do elements of equal keys maintain their relative order?), adaptibility (does partially sorted data change the algortihms efficiency?), and internal/external sorting (is the data entirely sorted in memory?).

**Bubble Sort**

The key to this sorting algorithm is understanding the concept of 'bubbling'. Each iteration of this algorithm bubbles the largest value to the back of the array. Repeating this again would then result the second largest value to back, then the third, and so on. After n iterations, the entire list would be sorted from the back to front. 

To bubble the largest value to the back of the list, we must compare each adjacent pair in the list and swap them if they are not in ascending order. One iteration takes O(n) time and sorts one value from the back. Thus, n iterations take O(n*n) time and sorts all n values from the back.

```python
def bubble_sort(arr):
    for _ in arr:
        for i in range(len(arr)-1):
            if arr[i] > arr[i+1]:
                arr[i], arr[i+1] = arr[i+1], arr[i]
    
    return arr

nums = [4,5,6,99,6,2,0,3,7,8]
print(bubble_sort(nums))
```

**Insertion Sort**

While bubble sort bubbles the maximum values out of the unsorted section, insertion sort bubbles elements into the sorted section. Bubble sort already knows the position it wants to fill, it just needs to find the value that belongs there. Insertion sort knows the value it wants to move, it just needs to find the position it belongs. 

To acheive this, insertion sort holds it's current value and shifts all the value before it forwards until the correct position is found. This sorts the list starting from the front of the list.

```python
def insertion_sort(arr):
	for i in range(1, len(arr)): 
		key = arr[i] 
		j = i-1
		while j >= 0 and key < arr[j]: 
			arr[j+1] = arr[j] 
			j -= 1
		arr[j+1] = key
	return arr

nums = [4,5,6,99,6,2,0,3,7,8]
print(insertion_sort(nums))
```

**Selection Sort**

Selection sort follows a very natural way of sorting items. It selects the smallest value in the unsorted list and swaps them to the front. This will sort the list from the front to back.

The processes of finding the minumum value in a list would mean looking through every value in the list. One iteration takes O(n) time and sorts one value. Thus, n iterations will take O(n*n) time and sorts all n values.

```python
def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

# Selection Sort (using min function)
def selection_sort_variation(arr):
    for i in range(len(arr)):
        min_index = min(range(i, len(arr)), key=lambda x: arr[x])
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
  
nums = [4,5,6,99,6,2,0,3,7,8]
print('Selection Sort: ', selection_sort(nums))
print('Selection Sort (using min function): ', selection_sort_variation(nums))
```

## Hash Structures 

Hash structure is a data structure that maps keys to values. A hash function takes an input key and generates mostly unique outputs (hash code) that acts as an index in memory. This means finding the key in memory is an O(1) operation, since it just has to compute the index from the key. However, this index does not contain the key's value. It is a bucket that holds a linked list that chains keys that produced the same hash code (forming a collision). In this case, the linked list would have to be traversed to find the exact key, making it an O(n) operation in instances of a collision. Other solutions to collisions include linear or quadratic probing to find the next buckets in a predefined sequence.

### Find Duplicates

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

```python
def containsDuplicate(self, nums: list[int]) -> bool:
    dup = set()
    
    for num in nums:
        if num in dup:
            return True
        dup.add(num)
    
    return False

def containsDuplicate(self, nums: list[int]) -> bool:
    dup = {}
    
    for num in nums:
        dup[num] = dup.get(num, 0) + 1
        if dup[num] >= 2:
            return True
            
    return False
```

### Valid Anagrams

Given two strings s and t, return true if t is an anagram of s, and false otherwise.
An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

```python
def isAnagram(self, s: str, t: str) -> bool:
    s_freq = {}
    t_freq = {}

    if len(s) != len(t):
        return False

    for ch in s:
        s_freq[ch] = s_freq.get(ch, 0) + 1
    
    for ch in t:
        t_freq[ch] = t_freq.get(ch, 0) + 1

    for s_ch,s_ch_Freq in s_freq.items():
        if not (s_ch in t_freq) or t_freq[s_ch] != s_ch_Freq:
            return False
    
    return True
```

### Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
You may assume that each input would have exactly one solution, and you may not use the same element twice.
You can return the answer in any order.

```python
def twoSum(self, nums: list[int], target: int) -> list[int]:
    comp_indices = {}

    for i in range(len(nums)):
        comp = target-nums[i]
        if comp in comp_indices:
            return [i,comp_indices.get(comp)]
        comp_indices[nums[i]] = i

    return []
```

## Linked Lists

A linked list is a data structure that represents a list of elements using nodes. Each node contains a value and a reference to the following node. These nodes are not contiguously stored in memory like arrays. Instead they are stored randomly and linked to each other through reference pointers. This allows them to have dynamic sizing without the O(n) time complexity dynamic arrays face when copying their data to larger contiguous memory slots because their current set filled up. However, linked lists lose the benefit of indexing, making accessing elements an O(n) operation instead of an O(1) one.

Certain variations like doubly linked lists also have a reference to the node prior to the current node. Other variations, such as a circular linked lists, point the tail (last node in the list) to the head (first node in the list).

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def _get_vals(self, res=''):
        if self.next is None:
            res += str(self.val)
            return res
        else:
            res += str(self.val) + '->'
            return self.next._get_vals(res)
    
    def __str__(self):
        return self._get_vals()
    

linked_list = ListNode(1,ListNode(2,ListNode(3,ListNode(4))))
print(linked_list)
```

### Insert a Target Element in a Sorted Linked List

### Find Kth Element from the End a Linked List in One Pass

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def _get_vals(self, res=''):
        if self.next is None:
            res += str(self.val)
            return res
        else:
            res += str(self.val) + '->'
            return self.next._get_vals(res)
    
    def __str__(self):
        return self._get_vals()
    

linked_list = ListNode(1,ListNode(2,ListNode(3,ListNode(4))))
print(linked_list)

def get_kth_elem(head: ListNode, k) -> int:

    if head is None:
        return None

    slow_ptr, fast_ptr = head, head

    for _ in range(k):
        if fast_ptr:
            fast_ptr = fast_ptr.next
        else:
            return -1

    while fast_ptr:
        slow_ptr = slow_ptr.next
        fast_ptr = fast_ptr.next

    return slow_ptr.val
    
print(get_kth_elem(linked_list, 3)) 
```

### Reverse a Linked List

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def _get_vals(self, res=''):
        if self.next is None:
            res += str(self.val)
            return res
        else:
            res += str(self.val) + '->'
            return self.next._get_vals(res)
    
    def __str__(self):
        return self._get_vals()
    

linked_list = ListNode(1,ListNode(2,ListNode(3,ListNode(4))))
print(linked_list)

def reverse_iterative(head: ListNode) -> ListNode:
    prev = None
    curr = head

    while curr:
        curr.next, prev, curr = prev, curr, curr.next
    return prev

def reverse_recursive_1(head: ListNode) -> ListNode:
    if head is None:
        return head
    if head.next is None:
        return head
    new_node = reverse_recursive_1(head.next)

    head.next.next = head
    head.next = None

    return new_node

def reverse_recursive_2(curr: ListNode, prev: ListNode = None) -> ListNode:
    next = curr.next
    curr.next = prev

    if next is None:
        return curr
    else:
        return reverse_recursive_2(next, curr)
    

linked_list = reverse_iterative(linked_list)
print(linked_list)
linked_list = reverse_recursive_1(linked_list)
print(linked_list)
linked_list = reverse_recursive_2(linked_list)
print(linked_list)
```

## Binary Trees

Binary Trees are hierarchical data structures. More specifically, it is a tree structure where each node at most two child nodes.

Tree terminologies:
1. Root (the topmost node)
2. Edge (the connection between two nodes)
3. Parent/child (parent nodes branch out to children nodes)
4. Leaf (a node with no children), sibling (nodes that share a parent)
5. Subtree (a tree of a child)
6. Height (number of edges from leaf node to another node).

Variations of binary trees: 
1. Binary search tree: every node's left subtree is smaller than itself and right subtree is greater than itself
2. Balanced binary tree: a tree whose heights of the left and right subtree do not differ by more than one (AVL, red-black trees..)
3. Complete binary tree: a tree whose every level is filled but the last one, which is filled from left to right (binary heaps)
4. Full binary trees: a tree who has either 0 or 2 children
5. Perfect binary tree: a tree which is both complete and full

```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def __str__(self):
        res = ''
        lines, *_ = self._display_aux()
        for line in lines:
            res += line + '\n'
        return res

    def _display_aux(self): # just for visual purposes (from)
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.val
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.left._display_aux()
        right, m, q, y = self.right._display_aux()
        s = '%s' % self.val
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

tree = TreeNode(1, TreeNode(2), TreeNode(3))
print(tree)
```

### Subsets
Given an integer array nums of unique elements, return all possible 
subsets (the power set).

The solution set must not contain duplicate subsets. Return the solution in any order.

Time & space complexity: O((2^n)*n)

```python
def subsets(nums: list[int]) -> list[list[int]]:
        res = []

        def dfs(i=0, candidates=[]):
            if i == len(nums):
                res.append(candidates[:])
                return 
            
            # exclude 
            dfs(i+1, candidates)

            # include
            candidates.append(nums[i])
            dfs(i+1, candidates)
            candidates.pop()

        
        dfs()
        return res
```

### Permutations
Time & Space Complexity: O(n!*n)

```python
def permute(nums: list[int]) -> list[list[int]]:
        res = []

        def dfs(i=0):
            if i == len(nums):
                res.append(nums[:])
                return

            for j in range(i,len(nums)):
                nums[i], nums[j] = nums[j], nums[i]
                dfs(i+1)
                nums[j], nums[i] = nums[i], nums[j]

        dfs()
        return res
```

### Combinations

Time & space complexity: O(n!*n)

```python
def combine(self, n: int, k: int) -> list[list[int]]:
        nums = [i+1 for i in range(n)]
        res = []

        def dfs(j = 0, candidates=[]):
            if len(candidates) == k:
                res.append(candidates[:])
                return

            for i in range(j,len(nums)):
                candidates.append(nums[i])
                dfs(i+1, candidates)
                candidates.pop()
        
        dfs()
        return res
```

```python
def dfs(adjacency_list):
    visited = set()

    def dfs_helper(node):
        visited.add(node)
        if node is None:
            return

        neighbors = adjacency_list.get(node)

        for next_node in neighbors:
            if next_node not in visited:
                dfs_helper(next_node)
        
        for node in adjacency_list.keys():
            if node not in visited:
                dfs_helper(next_node)
```

### Maximum Subarray (Kadane's Algorithm)

A brute force solution, O(n*n) time, would involve a nested for loop iterating over every possible contiguous subarray. A more efficient solution involves reducing that search space such that only the top few maximum subarrays are searched. 

First iterate over the list. At each iteration, decide what makes the biggest subarray: max_sum + curr_value or curr_value? If the current value is bigger without the baggage of the max sum calculated so far, then it is optimal for it to be it's own subarray. This will always ensure you know when to start building a new subarray. In other words it gives you the local maximum subarray relative to it's neighboring/adjacent subarray. 

However, we want the global maximum subarray. For that, we need another function checking if the local maximum is the biggest local maximum seen thus far. 

This has an O(n) time complexity.

```python
def maxSubArray(nums: list[int]) -> int:
        max_sum = float('-inf')
        curr_sum = 0
        for i in range(0,len(nums)):
            curr_sum = max(nums[i], curr_sum + nums[i])
            max_sum = max(max_sum, curr_sum)
        return max_sum

print(maxSubArray([10,-100,5,-2,6]))
```

### Permutation in String (Static Sliding Window)

Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.

*This is an unoptimized solution (track 'matches' for performant algorithm)

```python
def checkInclusion(s1: str, s2: str) -> bool:
        if len(s1) > len(s2):
            return False

        # hash
        s1_letters = {}
        s2_letters = {}
        for i in range(len(s1)):
            s1_letters[s1[i]] = s1_letters.get(s1[i],0) + 1
            s2_letters[s2[i]] = s2_letters.get(s2[i],0) + 1
            
        # anagram check
        def is_anagram():           
            for ch in s1:
                if ch not in s2_letters or s1_letters[ch] != s2_letters[ch]:
                    return False
            return True

        # sliding window
        left = 0
        for right in range(len(s1), len(s2)+1):
            if(is_anagram()):
                return True
            
            s2_letters[s2[left]] -= 1
            if right < len(s2):
                s2_letters[s2[right]] = s2_letters.get(s2[right],0) + 1
            left += 1
            
        return False
```

### Best Time to Buy and Sell Stocks (Dynamic Sliding Window)

You are given an array prices where prices[i] is the price of a given stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

Similar to Kadane's algorithm, our approach here is to find all the local maximum profits starting from all the local minimum values. We can then find which is the global maximum.

```python
def maxProfit(prices: list[int]) -> int:
    max_profit = 0
    min_buy = prices[0]
    
    left = 0
    for right in range(len(prices)):
        while prices[left] != min_buy:
            left += 1

        min_buy = min(min_buy, prices[right])
        max_profit = max(max_profit, prices[right]-prices[left])
    
    return max_profit
```

### Longest Substring Without Repeating Characters (Dynamic Sliding Window)

```python
def lengthOfLongestSubstring(s: str) -> int:
    max_len = 0
    seen = set()
    
    left = 0
    for right in range(len(s)):
        while s[right] in seen:
            seen.remove(s[left])
            left += 1
            
        seen.add(s[right])
        max_len = max(max_len, right - left + 1)
    
    return max_len
  
```

### Longest Repeating Character Replacement (Dynanmic Sliding Window)

You are given a string s and an integer k. You can choose any character of the string and change it to any other uppercase English character. You can perform this operation at most k times.

Return the length of the longest substring containing the same letter you can get after performing the above operations.

```python
def characterReplacement(self, s: str, k: int) -> int:
    max_len = 0
    freq = {}

    left = 0
    for right in range(len(s)):
        freq[s[right]] = freq.get(s[right], 0) + 1

        while (right-left+1) - max(freq.values()) > k:
            freq[s[left]] -= 1
            left += 1

        max_len = max(max_len, right-left+1)
        
    return max_len
```

### Reverse String (Static Two Pointer)

```python
def reverseString(s: list[str]) -> None:
    left = 0
    right = len(s) - 1

    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1
    
    return s
```

### Valid Palindrome (Dynamic Two Pointer)

```python
def isPalindrome(s: str) -> bool:
    left = 0
    right = len(s)-1

    def is_alphanum(ch):
        is_alpha = ord('a') <= ord(ch.lower()) <= ord('z') 
        is_digit = ord('0') <= ord(ch) <= ord('9')

        return is_alpha or is_digit

    while left < right:
        while left < right and not is_alphanum(s[left]):
            left += 1
        while right > left and not is_alphanum(s[right]):
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
    
    return True
```
