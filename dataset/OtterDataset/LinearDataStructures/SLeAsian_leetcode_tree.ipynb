All below are part of DFS:
- Pre-order: root, left ,right subtree
- In-order: left, root ,right subtree
- Post-order: left, right ,root subtree

### DFS and BFS time complexity: O(n)
- Because this is tree traversal, we must touch every node, making this O(n) where n is the number of nodes in the tree.

### BFS space complexity: O(n)
- BFS will have to store at least an entire level of the tree in the queue (sample queue implementation). With a perfect fully balanced binary tree, this would be (n/2 + 1) nodes (the very last level). Best Case (in this context), the tree is severely unbalanced and contains only 1 element at each level and the space complexity is O(1). Worst Case would be storing (n - 1) nodes with a fairly useless N-ary tree where all but the root node are located at the second level.

### DFS space complexity: O(d)
- Regardless of the implementation (recursive or iterative), the stack (implicit or explicit) will contain d nodes, where d is the maximum depth of the tree. With a balanced tree, this would be (log n) nodes. Worst Case for DFS will be the best case for BFS, and the Best Case for DFS will be the worst case for BFS.

# 104. Maximum Depth of Binary Tree
<a id='max_depth_binary_tree'></a>
https://leetcode.com/problems/maximum-depth-of-binary-tree/

intuition:
- https://www.youtube.com/watch?v=hTM3phVI6YQ (bfs and iterative/dfs)
- https://leetcode.com/problems/maximum-depth-of-binary-tree/discuss/359949/Python-recursive-and-iterative-solution (recursion/dfs and level order traversal/bfs)

recursion/dfs:
- recursion will hit base case for each branch first before moving on to the next one, causing it to be dfs.
- Max comparison at each branch, as base case is reached each path will start to be added up from bottom up.
- With comparison at each path when coming back up, each branch's max depth is found and eventually reach one max value.

iterative dfs:
- Same as above, comapre max at each branch.
- Use stack, first in last out to simulate depth first search.
- Need to keep track of depth for each node, so comparison is possible.

iterative (level-order-traversal, bfs):
1. Have a queue to keep track of which node to visit in order, a count, and the number of node left at each level.
2. Start with queue of length 1 (root) and traverse each level. At each node add children nodes to visit at the end of queue.
3. At the end of each level, update number of nodes left at new level and update count.

time complexity:
- recursion/dfs:
    - O(n), every node is visited.

- iterative (level-order-traversal, bfs):
    - O(n), every node is visited.

space:
- recursion/dfs:
    - logn on average for a balanced tree, 
    - O(n) worst case, degenerate tree (looks like linked list, only 1 child per node), beacuse the stack would be n length.

- iterative (level-order-traversal, bfs):
    - big theta(1) best case (degenerate tree looks like linked list, only 1 child per node), queue store one item at each level then update immediately.
    - O(n/2-1) = O(n) worse case, the max leaf width of a given level, because that's when the queue is most populated.

```python
#recursion/bfs
def maxDepth(self, root):
    if root is None:
        return 0
    return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1
```

```python
#iterative dfs
def maxDepth(root):
    stack = [[root,1]]
    res = 0
    
    while stack:
        node, depth = stack.pop()
        
        if node:
            res = max(res, depth)
            stack.append([node.left, depth + 1])
            stack.append([node.right, depth + 1])
    return res
```

```python
#iterative (level-order-traversal, bfs)
from collections import deque
#using deque for faster left pop, otherwise python list would work but each pop at left will need O(n) operation.
def maxDepth(root):
    if not root:
        return 0
    q = deque([root])
    level_count = 0
    num_node_level = 1
    while q:
        node = q.popleft()
        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)
        num_node_level -= 1
        if num_node_level == 0:
            level_count += 1
            num_node_level = len(q)
    return level_count
```

# 100. Same Tree
<a id='same_tree'></a>
https://leetcode.com/problems/same-tree/description/

intuition:
figured out by myself!

recursion/dfs:
- make sure know base case

iterative dfs:
- stack for first in last out

time complexity:
- recursion/dfs:
    - O(n), every node is visited.

- iterative (bfs):
    - O(n), every node is visited.

space:
- recursion/dfs:
    - logn on average for a balanced tree, 
    - O(n) worst case, degenerate tree (looks like linked list, only 1 child per node), beacuse the stack would be n length.

- iterative (bfs):
    - big theta(1) best case (degenerate tree looks like linked list, only 1 child per node), queue store one item at each level then update immediately.
    - O(n/2-1) = O(n) worse case, the max leaf width of a given level, because that's when the queue is most populated.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

#dfs recursion
def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
    def dfs(p, q):
        if p == None and q == None:
            return True
        if ((p == None or q == None) or
        (p.val != q.val)):
            return False
        final_match = all([dfs(p.left, q.left)
        ,dfs(p.right, q.right)])
        return final_match
    return dfs(p,q)

#dfs stack
def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
    stack = [(p,q)]
    while stack:
        left_t, right_t = stack.pop()
        if (left_t == None and right_t == None):
            continue
        if ((left_t == None and right_t != None) or
            (left_t != None and right_t == None) or
            (left_t.val != right_t.val)):
            return False

        stack.append((left_t.left, right_t.left))
        stack.append((left_t.right, right_t.right))
    return True
```

# 226. Invert Binary Tree
<a id='invert_binary_tree'></a>
https://leetcode.com/problems/invert-binary-tree/description/

intuition:
invert a tree, so switch invert left and right node of each layer BEFORE going to the next one. This will alternate the left and right subtree. Continously invert the layers below and it will be an inverted tree.

iterative dfs:
- stack for first in last out

time complexity:

- iterative (dfs):
    - O(n), every node is visited.

space complexity:

- iterative (dfs):
    - O(widest layer) worst case

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    #dfs stack
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        stack = [root]
        while stack:
            cur = stack.pop()
            if cur:
                cur.left, cur.right = cur.right, cur.left
                stack.extend([cur.left, cur.right])

        return root
```

# 102. Binary Tree Level Order Traversal
<a id='binary_tree_level_order_traversal'></a>
https://leetcode.com/problems/binary-tree-level-order-traversal/description/

solved by myself!

intuition:
level order traversal, left to right.
1. immediately thought of bfs, but how to get it to track each level?
2. notice branch grow like 1,2,4,8... (2 times of previous layer) and null branch can still be accounted.
3. keep track of branch number of each layer, minus one if null branch.

Figured out later, can just account how many branches for next with for loop.

time complexity: O(n), every node is visited.

space complexity: O(size of return array + size of queue) -> O(2N) worst case

```python
def levelOrder(root):
    #my solution
    lst = []
    
    if root:
        queue = [root]
        cur_lst = []
        branch_count = 1 #track how many branches
        cur_count = 1 #track current branch left to go

        while queue:
            cur_count -= 1
            cur = queue.pop(0) #not efficient O(n), use deque is better
            
            if cur:
                cur_lst.append(cur.val)
                queue.append(cur.left)
                queue.append(cur.right)
            else:
                branch_count -= 1

            if cur_count == 0:
                if cur_lst:
                    lst.append(cur_lst)
                branch_count *= 2
                cur_count = branch_count
                cur_lst = []
    return lst

from collections import deque
class Solution:
    def levelOrder(self, root):
        if not root: return []
        queue, res = deque([root]), []
        
        while queue:
            cur_level, size = [], len(queue)
            for i in range(size):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
                cur_level.append(node.val)
            res.append(cur_level)
        return res
```

# 572. Subtree of Another Tree
<a id='subtree_of_another_tree'></a>
https://leetcode.com/problems/subtree-of-another-tree/description/

intuition:

https://www.youtube.com/watch?v=E36O5SWp-LE

Solution is as intuitive as it sounds, check subRoot for each node and see if it matches. However, deployment is a bit harder.
- Split question into sameTree that does the tree matching at current node and iterating through the tree.

time complexity: O(nm) n is the number of nodes in root and m is number of node in subRoot

space complexity: O(1)

```python
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        if not subRoot: return True
        if not root and subRoot: return False

        if self.sameTree(root,subRoot):
            return True

        return (self.isSubtree(root.left,subRoot) or
                self.isSubtree(root.right,subRoot))

    def sameTree(self,a,b):
        if not a and not b:
            return True
        if a and b and a.val == b.val:
            return (self.sameTree(a.left, b.left) and
                    self.sameTree(a.right, b.right))
        return False
```

# 105. Construct Binary Tree from Preorder and Inorder Traversal
<a id='contruct_binary_tree_from_preorder_and_inorder_traversal'></a>
https://leetcode.com/problems/construct-binary-tree-from-preorder-and-inorder-traversal/description/

Pre-order:
- root, left, right subtree

In-order:
- left, root, right subtree

intuition:

https://www.youtube.com/watch?v=ihj4IQGZ2zc

- Notice first node for pre-order is always root.
- In-order's list left and right is split by root.
- Iterate down recursively and use pre-order to get each node at current until null.

time complexity: O(n) iterates through all the node

space complexity: O(1) if not including the answer

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        if not preorder or not inorder:
            return None
        
        root = TreeNode(preorder[0])
        mid_for_in_order = inorder.index(preorder[0])

        root.left = self.buildTree(preorder[1:mid_for_in_order + 1], inorder[:mid_for_in_order])
        root.right = self.buildTree(preorder[mid_for_in_order + 1:], inorder[mid_for_in_order + 1:])

        return root
```


# 235. Lowest Common Ancestor of a Binary Search Tree
<a id='lowest_common_ancestor_bst'></a>
https://leetcode.com/problems/lowest-common-ancestor-of-a-binary-search-tree/description/

intuition:

https://www.youtube.com/watch?v=gs2LMfuOR9k

1. Binary search tree, left subtree ALWAYS smaller, right ALWAYS bigger, assuming no duplicates.
2. This means if starting to traverse and cur node is between p and q, IT IS the lowest common ancestor, because the point where p and q are split, p is in left and q is in right subtree! They will never have common ancestor again.
3. Remember ancestor can be itself!

time complexity: O(logn) One node per level

space complexity: O(1)

```python
def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
    while root:
        if root.val < p.val and root.val < q.val:
            root = root.right
        elif root.val > p.val and root.val > q.val:
            root = root.left
        else:
            return root
```

```python
class Solution:
    # this solution works for even NON BST, works on Binary tree!
    # Time: O(n)
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        ans = self.helper(root, p, q)
        return ans

    def helper(self, root , p, q):
        if root:
            l = self.helper(root.left, p ,q)
            r = self.helper(root.right, p ,q)
            # if found node, return the node
            if str(type(r)) == "<class 'precompiled.treenode.TreeNode'>":
                return r
            if str(type(l)) == "<class 'precompiled.treenode.TreeNode'>":
                return l
            #traverse down and if found a node, return True and pass it up the tree
            if all([l, root.val in (p.val, q.val)]):
                return root
            if all([r, root.val in (p.val, q.val)]):
                return root
            if all([l,r]):
                return root
            if any([l,r,root.val in (p.val, q.val)]):
                return True
        return False
```

# 98. Validate Binary Search Tree
<a id='validate_binary_search_tree'></a>
https://leetcode.com/problems/validate-binary-search-tree/description/

intuition:
https://www.youtube.com/watch?v=s6ATEkipzow

- Binary search tree, left subtree ALWAYS smaller, right ALWAYS bigger, assuming no duplicates.
1. Initial approach, can brute force and check every node on left tree (are smaller) and right tree (are bigger). N nodes and each check n times, O(N^2), bad.
2. Think in terms of boundaries. Root node is bounded by -inf and inf. If go to left sub-tree, current node is bounded by the right ancestor, and vice versa. Remember to get the previous boundaries from ancestor for the side that is not updated!

<img src="photos/tree/validate_binary_search_tree.jpg" width="1000">

time complexity: O(n), 2N comparison for each node

space complexity: O(1)

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        def dfs(root, left, right):
            if not root: #no tree is good
                return True
            if not(left < root.val and root.val < right):
                return False
            return (dfs(root.left, left, root.val)  #move left, update right bound, leave the rest the same
                and dfs(root.right, root.val, right)) #move right, update left bound, leave the rest the same

        return dfs(root, float("-inf"), float("inf"))
```

# 230. Kth Smallest Element in a BST
<a id='kth_smallest_element_in_a_bst'></a>
https://leetcode.com/problems/kth-smallest-element-in-a-bst/description/

intuition:
https://www.youtube.com/watch?v=5LUXSvjmGCw

- Binary search tree, left subtree ALWAYS smaller, right ALWAYS bigger, assuming no duplicates.
Two approaches:
- Recursive: In order traversal, create ordered list. simply select kth index.

time complexity: O(n)

space complexity: O(n)

- Iterative: In order traversal, stop when count from bottom up is at the right count.

time complexity: O(n), if tree all left or right

space complexity: O(n), if tree all left or right

```python
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        def inorder(root):
            if not root:
                return []
            return inorder(root.left) + [root.val] + inorder(root.right)
        
        return inorder(root)[k - 1]
```
