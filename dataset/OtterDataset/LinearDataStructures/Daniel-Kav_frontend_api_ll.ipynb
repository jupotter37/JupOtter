insertAtPosition Function:
If the position is 0, the new node becomes the head, and its next points to the old head.
Otherwise, we traverse the list to find the node just before the insertion point. Once we find it, we adjust the pointers: the new node points to the next node in sequence, and the current node points to the new node.

```python
# Python program to merge two sorted linked lists
# using dummy nodes

class Node:
    def __init__(self, new_data):
        self.data = new_data
        self.next = None

# Function that takes two lists sorted in increasing order,
# and splices their nodes together to make one big sorted
# list which is returned.
def merge_sorted_list(first, second):

    # A dummy first node to hang the result on
    dummy = Node(-1)

    # Tail points to the last result node
    tail = dummy

    # So tail.next is the place to add new nodes to the
    # result
    while True:
        if first is None:

            # If either list runs out, use the other list
            tail.next = second
            break
        elif second is None:
            tail.next = first
            break
        if first.data <= second.data:

            # Move the front node from 'first' to the result
            # list
            new_node = first
            first = first.next
            new_node.next = tail.next
            tail.next = new_node
        else:

            # Move the front node from 'second' to the result
            # list
            new_node = second
            second = second.next
            new_node.next = tail.next
            tail.next = new_node

        tail = tail.next
    return dummy.next

# Driver code

# Create a hard-coded linked list:
# 2 -> 4 -> 8 -> 9
first = Node(2)
first.next = Node(4)
first.next.next = Node(8)
first.next.next.next = Node(9)

# Create another hard-coded linked list:
# 1 -> 3 -> 8 -> 10
second = Node(1)
second.next = Node(3)
second.next.next = Node(8)
second.next.next.next = Node(10)

merged_list = merge_sorted_list(first, second)

temp = merged_list
print("Merged Link List is:")
while temp:
        print(temp.data, end=" ")
        temp = temp.next
```

```python
#find kth element from the end of a linked list 

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def getKthFromEnd(head, k):
    fast = head
    slow = head

    # Move fast pointer k steps ahead
    for _ in range(k):
        if fast is None:  # If k is greater than the number of nodes
            return -1
        fast = fast.next

    # Move both pointers until fast reaches the end
    while fast:
        slow = slow.next
        fast = fast.next

    # slow is now at the kth node from the end
    return slow.data if slow else -1

# Helper function to print a linked list
def printList(head):
    while head:
        print(head.data, end=" -> ")
        head = head.next
    print("None")

# Test the function
if __name__ == "__main__":
    # Create a linked list: 1->2->3->4->5->6->7->8->9
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)
    head.next.next.next.next = Node(5)
    head.next.next.next.next.next = Node(6)
    head.next.next.next.next.next.next = Node(7)
    head.next.next.next.next.next.next.next = Node(8)
    head.next.next.next.next.next.next.next.next = Node(9)

    # Example 1
    k = 2
    print(f"Kth node from end when k={k}: {getKthFromEnd(head, k)}")  # Output: 8

    # Example 2
    head2 = Node(10)
    head2.next = Node(5)
    head2.next.next = Node(100)
    head2.next.next.next = Node(5)
    k = 5
    print(f"Kth node from end when k={k}: {getKthFromEnd(head2, k)}")  # Output: -1
```

```python
#removing duplicates from a sorted linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeDuplicates(head):
    if not head:
        return None

    current = head

    while current and current.next:
        if current.data == current.next.data:
            # Skip the duplicate node
            current.next = current.next.next
        else:
            # Move to the next node
            current = current.next

    return head

# Helper function to print the linked list
def printList(head):
    while head:
        print(head.data, end=" -> ")
        head = head.next
    print("None")

# Test the function
if __name__ == "__main__":
    # Create a sorted linked list: 1->1->2->3->3
    head = Node(1)
    head.next = Node(1)
    head.next.next = Node(2)
    head.next.next.next = Node(3)
    head.next.next.next.next = Node(3)

    print("Original list:")
    printList(head)

    # Remove duplicates
    head = removeDuplicates(head)

    print("List after removing duplicates:")
    printList(head)
```

```python
#doubly linked lists
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

class doublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            new_node.prev = None
            self.head = new_node
        else:
            curr = self.head
            while curr.next:
                curr = curr.next
            curr.next = new_node
            new_node.next = None
            new_node.prev = curr
        return self.head
    
    def prepend(self, data):
        new_node = Node(data)
        if not self.head:
            new_node.prev = None
            self.head = new_node
        else:
            self.head.prev = new_node
            new_node.next = self.head
            new_node.prev = None
            self.head = new_node
        return self.head
    
    def add_after_node(self,key,data):
        new_node = Node(data)
        curr = self.head
        while curr:
            if curr.next is None and curr.data == key:
                self.append(data)
                break
            if curr.data == key:
                temp = curr.next
                curr.next = new_node
                new_node.next = temp
                temp.prev = new_node
                new_node.prev = curr
            curr = curr.next
            
        if curr is None:
            print(f'{key} out of bounds')
        return  self.head
    
    def add_b4_node(self,key,data):
        new_node = Node(data)
        curr = self.head
        while curr:
            if curr.prev is None and curr.data == key:
                self.prepend(data)
                break
            elif curr.data == key:
                temp = curr.prev
                temp.next = new_node
                new_node.next = curr
                curr.prev = new_node
                new_node.prev = temp
            curr = curr.next
        return self.head
    
    def delete(self,key):
        curr = self.head
        while curr:
            if curr.data == key and curr == self.head:
                #if you  are deleting at the head and ther is a next element
                if curr.next:
                    nxt = curr.next
                    curr.next = None
                    curr = None
                    nxt.prev = None
                    self.head = nxt
                    return
                #if you  are deleting at the head and there is no  next element
                else:
                    curr = None
                    self.head = None
                    return
            if curr.data == key:
                # deleting in the doubly linkedlist
                if curr.next:
                    nxt = curr.next
                    prev = curr.prev
                    nxt.prev = prev
                    prev.next = nxt
                    curr.next = None
                    curr.prev = None
                    del curr
                    return
                # deleting at the end of a doubly linkedlist
                else:
                    prev = curr.prev
                    prev.next = None
                    curr.prev = None
                    curr = None
                    return
                
            curr = curr.next
        return self.head
    
    def print_dlist(self):
        curr = self.head
        while curr:
            print(f'{curr.data}', end='<->')
            curr = curr.next
        print('None')
dll = doublyLinkedList()
dll.append(2)
dll.append(3)
dll.append(7)
dll.add_after_node(3,5)
dll.add_b4_node(2,33)
dll.add_b4_node(5,9)
dll.delete(33)
dll.print_dlist()
    
```

```python
# Subarray range with given sum
# Given an unsorted array of integers arr[], and a target tar, determine the number of subarrays whose elements sum up to the target value.

def sub_arr_sum(arr,target):
    #create a dictionary to store the sums and their counts
    prefix_sums = {0: 1}

    # initialize a current sum and count
    current_sum = 0
    count = 0

    # loop through the array 
    for num in arr:
    #update the running sum
        current_sum += num

        # check if the (current sum - target)is in the prefix sums map
        #if it is inrement the count
        if (current_sum - target) in prefix_sums:
            count += prefix_sums[current_sum - target]
        
    
    # update the prefix sums map with the current sum
        if current_sum in prefix_sums:
            prefix_sums[current_sum ] += 1
        else:
            prefix_sums[current_sum] = 1
    return count

arr1 = [10, 2, -2, -20, 10]
target1 = -10
print(sub_arr_sum(arr1, target1))  # Output: 3

arr2 = [1, 4, 20, 3, 10, 5]
target2 = 33
print(sub_arr_sum(arr2, target2)) 


```

```python
# Given an array arr. Calculate the sum of lengths of contiguous subarrays having all distinct elements.
# Note: Return the answer with modulo 109+7.

def sumOfLengths(arr):
    MOD = 10**9 + 7
    n = len(arr)
    
    # To store the final result
    result = 0
    # Set to store distinct elements in the current window
    seen = set()
    
    # Initialize two pointers
    left = 0
    
    # Traverse the array with the right pointer
    for right in range(n):
        # If the element at arr[right] is already in the window, 
        # move the left pointer to remove duplicates
        while arr[right] in seen:
            seen.remove(arr[left])
            left += 1
        
        # Add the current element to the set
        seen.add(arr[right])
        
        # Calculate the number of valid subarrays ending at 'right'
        # and add it to the result
        result += (right - left + 1)
        result %= MOD
    
    return result

# Example usage:
arr1 = [1, 2, 3]
arr2 = [1]

print(sumOfLengths(arr1))  # Output: 10
print(sumOfLengths(arr2))  # Output: 1
```

```python
#Implementing a stack using a queue
from collections import deque
class Stack:
    def __init__(self):
        #initialize a queue
        self.q = deque()
    def push(self,data):
        self.q.append(data)
    def pop(self):
        for i in range(len(self.q) - 1):
            self.push(self.q.popleft())
        return self.q.popleft()
    def top(self):
        return self.q[-1]
    def isEmpty(self):
        return len(self.q) == 0
    def print(self):
        for i in self.q:
            print(i,end=' ')


st = Stack()

st.push(1)
st.push(2)
st.push(3)
st.pop()
print(st.top())
print(st.isEmpty())
st.print()
```

```python
# Given a permutation of some of the first natural numbers in an array arr[], determine if the array can be sorted in exactly two swaps. A swap can involve the same pair of indices twice.

# Return true if it is possible to sort the array with exactly two or zero swaps, otherwise return false.

def check_sorted(arr):
    n = len(arr)
    swap_count = 0 

    #attempt to place each element at its corect position
    for i in range(n):
        while arr[i] != i + 1:
            # perform swap with the correct position of arr[i] 
            arr[arr[i] - 1],arr[i] = arr[i],arr[arr[i] - 1]
            swap_count += 1

            #if the swaps exceed two return false
            if swap_count > 2:
                return False
    for i in range(n):
        if arr[i] != i + 1:
            return False
        
    return swap_count == 0 or swap_count == 2

print(check_sorted([4, 3, 2, 1]))  # Expected output: True
print(check_sorted([4, 3, 1, 2]))  # Expected output: False
print(check_sorted([1, 2, 3, 4]))  # Expected output: True
```

```python
#Merge strings alternately
def merge_str(word1,word2):
    min_length = min(len(word1),len(word2))
    res = []
    for i in range(min_length):
        res.append(word1[i])
        res.append(word2[i])
    
    res.append(word1[min_length:])
    res.append(word2[min_length:])

    return ''.join(res)

word1 = 'abz'
word2 = 'cdf'

print(merge_str(word1,word2))
```

```python
class ListNode:
    def __init__(self, data=0, next=None):
        self.data = data
        self.next = next

def reverse_linked_list(head):
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev  # New head after reversing

def add_one(head):
    # Step 1: Reverse the linked list
    head = reverse_linked_list(head)

    # Step 2: Add one to the reversed linked list
    current = head
    carry = 1  # We start with carry 1 because we want to add one

    while current and carry:
        current.data += carry
        if current.data == 10:  # If we have a carry from this digit
            current.data = 0  # Reset to 0 and carry forward
            carry = 1
        else:
            carry = 0  # No carry, we can stop adding
        current = current.next

    # Step 3: If there's still a carry left, we need to add a new node
    if carry:
        new_node = ListNode(1)
        new_node.next = head  # New node becomes the new head
        head = new_node

    # Step 4: Reverse the linked list back to original order
    return reverse_linked_list(head)  # Return the modified list

# Example usage:
def print_linked_list(head):
    current = head
    while current:
        print(current.data, end="")
        current = current.next
    print()

# Creating linked lists for the examples
# Linked List: 4 -> 5 -> 6
head1 = ListNode(4, ListNode(5, ListNode(6)))
result1 = add_one(head1)
print_linked_list(result1)  # Output: 457

# Linked List: 1 -> 2 -> 3
head2 = ListNode(1, ListNode(2, ListNode(3)))
result2 = add_one(head2)
print_linked_list(result2)  # Output: 124
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Solution:
    def quickSort(self, head):
        if not head or not head.next:
            return head  # Base case for empty list or single element
        
        # Recursive quicksort call with the head and tail of the list
        return self.quickSortRec(head, self.getTail(head))

    # Utility function to get the tail of the list
    def getTail(self, node):
        while node and node.next:
            node = node.next
        return node

    # Recursive quicksort function
    def quickSortRec(self, start, end):
        if start == end:
            return start  # Base case for single node
        
        lessHead = None
        pivot = None
        greaterHead = None

        # Partition the list around the pivot (first element as pivot)
        lessHead, pivot, greaterHead = self.partition(start, end)

        # Sort the left and right partitions
        lessHead = self.quickSortRec(lessHead, self.getTail(lessHead)) if lessHead else None
        greaterHead = self.quickSortRec(greaterHead, self.getTail(greaterHead)) if greaterHead else None

        # Concatenate less, pivot, and greater lists and return the new head
        return self.concatenate(lessHead, pivot, greaterHead)

    # Partition function using the first node as pivot
    def partition(self, start, end):
        pivot = start
        lessHead = None
        greaterHead = None
        
        current = start.next
        lessTail = None
        greaterTail = None

        # Partition nodes into less and greater lists
        while current != end.next:
            if current.data < pivot.data:
                if not lessHead:
                    lessHead = lessTail = current
                else:
                    lessTail.next = current
                    lessTail = current
            else:
                if not greaterHead:
                    greaterHead = greaterTail = current
                else:
                    greaterTail.next = current
                    greaterTail = current
            current = current.next

        # End both lists
        if lessTail:
            lessTail.next = None
        if greaterTail:
            greaterTail.next = None

        return lessHead, pivot, greaterHead

    # Concatenate less, pivot, and greater lists
    def concatenate(self, less, pivot, greater):
        head = less

        if less:
            tail = self.getTail(less)
            tail.next = pivot
        else:
            head = pivot

        pivot.next = greater
        return head
```
