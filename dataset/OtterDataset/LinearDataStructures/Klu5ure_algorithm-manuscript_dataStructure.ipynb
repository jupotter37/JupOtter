# 动态数组

```python
class ArrayList:

    def __init__(self):
        self.list = []

    def get(self, index: int) -> int:
        if index < 0 or index >= len(self.list):
            return -1
        return self.list[index]

    def addAtHead(self, val: int) -> None:
        self.list.insert(0, val)

    def addAtTail(self, val: int) -> None:
        self.list.append(val)

    def addAtIndex(self, index: int, val: int) -> None:
        if index < 0 or index > len(self.list):
            return -1
        self.list.insert(index, val)

    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= len(self.list):
            return -1
        self.list.pop(index)
```

```python
# 假如要自己实现一个动态数组，而不是用python自带的list
# 不能用list的方法，直接访问和修改list[i]
class ArrayList:

    def __init__(self):
        self.capacity = 10
        self.size = 0
        self.extend_ratio = 2
        self.list = [0] * self.capacity
        
    def get(self, index: int) -> int:
        if index < 0 or index >= self.size:
            return -1
        return self.list[index]

    def addAtHead(self, val: int) -> None:
        if self.size == self.capacity:
            self.extend_capacity()
        for i in range(self.size, 0, -1):
            self.list[i] = self.list[i - 1]
        self.list[0] = val
        self.size += 1
        

    def addAtTail(self, val: int) -> None:
        if self.size == self.capacity:
            self.extend_capacity()
        self.list[self.size] = val
        self.size += 1

    def addAtIndex(self, index: int, val: int) -> None:
        if index < 0 or index > self.size:
            return -1
        if self.size == self.capacity:
            self.extend_capacity()
        for i in range(self.size, index, -1):
            self.list[i] = self.list[i - 1]
        self.list[index] = val
        self.size += 1

    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= self.size:
            return -1
        for i in range(index, self.size - 1):
            self.list[i] = self.list[i + 1]
        self.list[self.size - 1] = 0
        self.size -= 1
        
    def extend_capacity(self):
        self.list = self.list + [0] * self.capacity * (self.extend_ratio - 1)
        self.capacity = len(self.list)
        
# list = ArrayList()
# list.addAtTail(2)
# list.addAtTail(4)
# list.addAtHead(1)
# list.addAtIndex(2,3)
# list.deleteAtIndex(3)
# print(list.list)
# list.size

list = ArrayList()
list.addAtHead(7)
list.addAtHead(2)
list.addAtHead(1)
list.addAtIndex(3,8)
print(list.get(3))
# list.deleteAtIndex(2)
# list.addAtHead(6)
# list.addAtTail(4)
# list.get(4)
list.list
```

# 链栈

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class MyStack:

    def __init__(self):
        self.top_node = None

    def push(self, x: int) -> None:
        node = Node(x)
        node.next = self.top_node
        self.top_node = node

    def pop(self) -> int:
        if self.empty():
            return -1
        node = self.top_node
        self.top_node = node.next
        return node.data

    def top(self) -> int:
        if self.empty():
            return -1
        return self.top_node.data

    def empty(self) -> bool:
        return self.top_node is None


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

# 队列栈

```python
# 用列表模拟队列，只能用pop(0)

class MyStack:

    def __init__(self):
        self.pushQueue = []
        self.popQueue = []
        

    def push(self, x: int) -> None:
        self.pushQueue.append(x)

    def pop(self) -> int:
        if not self.popQueue:
            while len(self.pushQueue) > 1:
                self.popQueue.append(self.pushQueue.pop(0))
        res = self.pushQueue.pop()
        self.pushQueue, self.popQueue = self.popQueue, self.pushQueue
        return res 

    def top(self) -> int:
        if not self.popQueue:
            while len(self.pushQueue) > 1:
                self.popQueue.append(self.pushQueue.pop(0))
        res = self.pushQueue[0]
        self.popQueue.append(self.pushQueue.pop())
        self.pushQueue, self.popQueue = self.popQueue, self.pushQueue
        return res


    def empty(self) -> bool:
        return not self.popQueue and not self.pushQueue



# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None

class MyQueue:

    def __init__(self):
        self.head = None 

    def push(self, x: int) -> None:
        node = self.head
        if not self.head:
            self.head = Node(x)
            return
        while node.next:
            node = node.next
        node.next = Node(x)

    def pop(self) -> int:
        node = self.head
        self.head = self.head.next
        return node.data


    def peek(self) -> int:
        return self.head.data


    def empty(self) -> bool:
        return self.head is None
    

    
    
    
# 上面的方式每次push都要遍历，所以用下面的，同时记录队首和队尾
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class MyQueue:

    def __init__(self):
        self.head = None
        self.tail = None

    def push(self, x: int) -> None:
        node = Node(x)
        if not self.head:
            self.head = node
            self.tail = node
            return
        self.tail.next = node
        self.tail = node

    def pop(self) -> int:
        node = self.head
        self.head = self.head.next
        return node.data


    def peek(self) -> int:
        return self.head.data

    def empty(self) -> bool:
        return not self.head


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

# 栈队列

```python
class MyQueue:

    def __init__(self):
        self.pushStack = []
        self.popStack = []

    def push(self, x: int) -> None:
        self.pushStack.append(x)

    def pop(self) -> int:
        if not self.popStack:
            while self.pushStack:
                self.popStack.append(self.pushStack.pop())
        return self.popStack.pop()

    def peek(self) -> int:
        if not self.popStack:
            while self.pushStack:
                self.popStack.append(self.pushStack.pop())
        return self.popStack[-1]

    def empty(self) -> bool:
        return not self.pushStack and not self.popStack
```

```python
# 合并k个有序链表，低效率
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = []
        head = ListNode()
        tail = head
        heap = [node for node in lists if node]
        while heap:
            heap.sort(key=lambda x : x.val)
            node = heap.pop(0)
            tail.next = node
            tail = tail.next
            if node.next:
                heap.append(node.next)
        return head.next
```

```python
# 使用自带的最小堆来解决合并k个有序链表问题

from heapq import *
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

ListNode.__lt__ = lambda a, b : a.val < b.val
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = []
        head = ListNode()
        tail = head
        heap = [node for node in lists if node]
        heapify(heap)
        while heap:
            node = heappop(heap)
            tail.next = node
            tail = tail.next
            if node.next:
                heappush(heap, node.next)
        return head.next
```

# BFS

```python
from collections import deque

class Solution:
    def minDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        return self.bfs(root)

    def bfs(self, root):
        q = deque()
        q.append(root)
        res = 0
        while q:
            res += 1
            for i in range(len(q)):
                node = q.popleft()
                if not node.left and not node.right:
                    return res
                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)
```

# DFS

```python
class Solution:
    def __init__(self):
        self.res = []

    def permute(self, nums: List[int]) -> List[List[int]]:
        self.dfs(nums, [], [False for _ in nums])
        return self.res

    def dfs(self, nums, path, used):
        if len(path) == len(nums):
            self.res.append(path[:])
            return
        for i in range(len(used)):
            if not used[i]:
                used[i] = True
                path.append(nums[i])
                self.dfs(nums, path, used)
                used[i] = False
                path.pop()
```
