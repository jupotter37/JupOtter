# Linked List

```python
class ListNode():
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
        
class LinkedList():
    def __init__(self):
        self.head = None
    
    def append(self,data):
        assert isinstance(data,int), "Data should be integer datatype"
        if self.head is None:
            self.head = ListNode(data,None)
            return
        temp = self.head
        while temp.next:            
            temp = temp.next
        temp.next = ListNode(data,None)
        
    def print(self):
        if self.head is None:
            print("Empty List!")
            return
        temp = self.head
        while temp:
            print(temp.data,end="-> ")
            temp = temp.next
        print("NULL")
        
    def length(self):
        c = 0
        temp = self.head
        while temp:
            c+=1
            temp = temp.next
        return c
    
    def popleft(self):
        if self.length() == 1:
            self.head = None
            return
        self.head = self.head.next
        
    def insert(self,ind,data):
        assert ind<=self.length()-1, "Index out of bound"
        if ind==self.length()-1:
            self.append(data)
            return
        elif ind==0:
            node = ListNode(data,self.head)
            self.head = node
        else:
            temp = self.head
            c = 0
            while temp:
                if c==ind:                    
                    temp.next = ListNode(data,temp.next)
                    return
                temp = temp.next     
                c+=1
        
    def remove_end(self):
        if self.length() == 1:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None
        
    def remove(self,ind):
        if ind == self.length()-1:
            self.remove_end()
            return
        elif ind==0:
            self.popleft()
            return
        temp = self.head
        c= 0
        while temp:
            if c==ind:
                temp.next = temp.next.next
            temp = temp.next
            c+=1
        
if __name__ == "__main__":
    ll = LinkedList()
    ll.append(25)
    ll.append(35)
    ll.append(21)
    ll.append(30)
    ll.append(56)
    ll.append(69)
    ll.print()
    ll.popleft()
    ll.print()
    ll.insert(4,33)
    ll.print()
    ll.remove_end()
    ll.print()
    ll.remove(1)
    ll.print()
```

# Queue

```python
from collections import deque
class Queue():
    def __init__(self):
        self.queue = deque()
    
    def enqueue(self,data):
        self.queue.append(data)
        
    def dequeue(self):
        self.queue.popleft()
    
    def print(self):
        print(self.queue)
        
if __name__ == "__main__":
    q = Queue()
    q.enqueue(56)
    q.enqueue(69)
    q.dequeue()
    q.print()
```

# Tree

```python
class TreeNode():
    def __init__(self,data):
        self.data = data
        self.children = []
        self.parent = None
    
    def add_child(self,child):
        child.parent= self
        self.children.append(child)
        
    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level+=1
            p = p.parent
        return level
    
    def print_tree(self):
        spaces = ' '*self.get_level()*2
        prefix = spaces + "|-" if self.parent else ""
        print(prefix+self.data)
        if self.children:
            for child in self.children:
                child.print_tree()
    
def build_tree():
    root = TreeNode("Electronics")
    laptop = TreeNode("Laptop")
    laptop.add_child(TreeNode("Macbook"))
    laptop.add_child(TreeNode("Asus"))
    laptop.add_child(TreeNode("Acer"))
    
    cell = TreeNode("Cell Phone")
    cell.add_child(TreeNode("Apple"))
    cell.add_child(TreeNode("Google"))
    cell.add_child(TreeNode("Samsung"))
    
    tv = TreeNode("TV")
    tv.add_child(TreeNode("Samsung"))
    tv.add_child(TreeNode("Oneplus"))
    tv.add_child(TreeNode("Sony"))
    
    root.add_child(laptop)
    root.add_child(tv)
    root.add_child(cell)
    
    root.print_tree()

if __name__=="__main__":
    build_tree()
```

# Binary Tree

```python
class Node():
    def __init__(self,data,left=None,right=None):
        self.data = data
        self.left = left
        self.right = right

class BinaryTree():
    def __init__(self,data):
        self.root = Node(data)
        
    def add_child(self,data):
        if data == self.root.data:
            return
        if data<self.root.data:
            if self.root.left:
                self.root.left.add_child(data)
            else:
                self.root.left = BinaryTree(data)
        else:
            if self.root.right:
                self.root.right.add_child(data)
            else:
                self.root.right= BinaryTree(data)
                
    def in_order(self):
        elements = []
        if self.root.left:
            elements+=self.root.left.in_order()
        elements.append(self.root.data)
        if self.root.right:
            elements+=self.root.right.in_order()
        return elements
    
    def post_order(self):
        elements = []
        if self.root.left:
            elements+=self.root.left.post_order()
        if self.root.right:
            elements+=self.root.right.post_order()
        elements.append(self.root.data)
        return elements
    
    def pre_order(self):
        elements = []
        elements.append(self.root.data)
        if self.root.left:
            elements+=self.root.left.pre_order()
        if self.root.right:
            elements+=self.root.right.pre_order()        
        return elements
    
    def search(self,data):
        if self.root.data == data:
            return True
        if data<self.root.data:
            if self.root.left:
                return self.root.left.search(data)
            return False
        
        if data>self.root.data:
            if self.root.right:
                return self.root.right.search(data)
            return False
        return False
    
    def max(self):
        if not self.root.right:
            return self.root.data
        return self.root.right.max()
        
    def min(self):
        if self.root.left:
            return self.root.left.min()
        return self.root.data
    
    def sum(self):
        m = self.root.data
        if self.root.left:
            m+=self.root.left.sum()
        if self.root.right:
            m+=self.root.right.sum()
        return m
    
    def delete(self,data):
        if not self.root:
            return None
        if self.root.data>data:
            if self.root.left:
                self.root.left = self.root.left.delete(data)
        elif self.root.data<data:
            if self.root.right:
                self.root.right = self.root.right.delete(data)
        else:
            if self.root.left is None and self.root.right is None:
                return None
            
            if self.root.left is None:
                return self.root.right
            
            if self.root.right is None:
                return self.root.left
            
            maxx = self.root.left.max()
            self.root.data = maxx
            self.root.left = self.root.left.delete(maxx)
            
        return self
    
def build_tree(arr):
    root = BinaryTree(arr[0])
    for i in range(1,len(arr)):
        root.add_child(arr[i])
    return root
        
if __name__ =="__main__":
    numbers = [17, 4, 1, 20, 9, 23, 18, 34]
    root = build_tree(numbers)
    print(f"In order: {root.in_order()}")
    print(f"Post order: {root.post_order()}")
    print(f"Pre order: {root.pre_order()}")
    print(root.search(40))
    print(f"Max: {root.max()}")
    print(f"Min: {root.min()}")
    print(f"Sum: {root.sum()}")
    root.delete(4)
    print(f"In order: {root.in_order()}")
    
```

# Graph

```python
class Graph():
    def __init__(self,edges):
        self.edges = edges
        self.graph = {}
        for src , dest in edges:
            if src in self.graph:
                self.graph[src].append(dest)
            else:
                self.graph[src] = [dest]
        
    def get_paths(self,src,dest,path=[]):
        path = path + [src]
        if src==dest:
            return [path]
        if src not in self.graph:
            return []
        paths = []
        for node in self.graph[src]:
            if node not in path:
                new_paths = self.get_paths(node,dest,path)
                for p in new_paths:
                    paths.append(p)
        return paths
    
    def shortest_path(self,src,dest,path=[]):
        path = path + [src]
        if src==dest:
            return path
        if src not in self.graph:
            return []
        shortest_path = None
        for node in self.graph[src]:
            if node not in path:
                temp_sp = self.shortest_path(node,dest,path)
                if temp_sp:
                    if shortest_path is None or len(temp_sp)<len(shortest_path):
                        shortest_path = temp_sp
        return shortest_path
if __name__ =="__main__":
    routes = [
        ("Mumbai","Pune"),
        ("Mumbai", "Surat"),
        ("Surat", "Bangaluru"),
        ("Pune","Hyderabad"),
        ("Pune","Mysuru"),
        ("Hyderabad","Bangaluru"),
        ("Hyderabad", "Chennai"),
        ("Mysuru", "Bangaluru"),
        ("Chennai", "Bangaluru")
        ]
    graph = Graph(routes)
    print(f'All paths: {graph.get_paths("Mumbai","Bangaluru")}')
    print(f'\nShortest path: {"-> ".join(graph.shortest_path("Mumbai","Bangaluru"))}')
```
