# 3.2 Stack Min: Correct
How would you design a stack which, in addition to push and pop, has a function min which returns the minimum element? Push, pop and min should all operate in 0(1) time.

## Answer 1
When you push an element, compare with element on top with peek. If element is less than top element then push, if element is greater than top element then pop top element, push greater element, then re push the old top element. 
This way whenever you pop an element from the list it is the min.

## Answer 2 after Hint 2

When you push an element, compare with element on top with peek. If element is less than top element then set new element to min in the stack (i.e have bool for each element in stack if min or not), if element is greater than top element then set min = false for newly pushed element 
To retrieve the min just grab element in stack that has min = true


## Answer 3 after Hint 3

When you push an element, compare with element on top with peek. If element is less than top element then also add data to new element that = min in the stack (i.e have variable for each element in stack of the latest min), if element is greater than top element then keep min = to last min for newly pushed element 
To retrieve the min just pop the top element and also look at the additional variable in the element that = the min 

```python
class Node:
    def __init__(self,value,min_value):
        self.value = value
        self.min_value = min_value

class Stack:

    def __init__(self):
        self.stack = []

    def push(self, value):
        '''Push element to the top of stack'''
        if (len(self.stack) < 1):
            min_value = value
            new_node = Node(value,min_value)
        else:
            if value < self.peek().value:
                min_value = value
            else:
                min_value = self.peek().min_value
            
            new_node = Node(value,min_value)
        
        if value not in self.stack:
            self.stack.append(new_node)
            return True
        else:
            print("Element already on top")
            return False
        
    def pop(self):
        if len(self.stack) <= 0:
            return ("No element in the Stack")
        else:
            return self.stack.pop()

    def peek(self): 
        '''Return top element of the stack'''
        return self.stack[-1]
    
    def find_min(self): 
        '''Return top element of the stack with min value'''
        if (len(self.stack) > 0): 
            print("Min: ", self.stack[-1].min_value)
    
    def printStack(self): 
        '''Print all elements of the stack'''
        for element in self.stack:
            print(element.value, " - min: ",element.min_value)
      
```

# 3.3 Stack of Plates: Correct
Imagine a (literal) stack of plates. If the stack gets too high, it might topple. Therefore, in real life, we would likely start a new stack when the previous stack exceeds some threshold. Implement a data structure SetOfStacks that mimics this. SetOfStacks should be composed of several stacks and should create a new stack once the previous one exceeds capacity. SetOfStacks.push() and SetOfStacks.pop() should behave identically to a single stack (that is, pop () should return the same values as it would if there were just a single stack).

FOLLOW UP

Implement a function popAt(int index) which performs a pop operation on a specific sub-stack.

```python
class SetOfStacks:
    
    def __init__(self, threshold):
        self.set_of_stacks = []
        self.stack = [] 
        self.set_of_stacks.append(self.stack)
        
        self.threshold = threshold
        
    
    def push(self,value):
        
        #if length of stack is too tall, create new stack and append to set of stacks
        if len(self.stack) >= self.threshold:
            self.stack = []
            self.stack.append(value)
            self.set_of_stacks.append(self.stack)
        #else just add value to current stack 
        else:
            self.stack.append(value)
            
    def pop(self):
        if len(self.set_of_stacks[-1]) > 0:
            last_element =  self.set_of_stacks[-1].pop()
            if len(self.set_of_stacks[-1]) == 0:
                self.set_of_stacks.pop()
            return last_element
        else:
            return self.set_of_stacks.pop()
    
    def popAt(self, index):
        if len(self.set_of_stacks[index]) > 0:
            last_element = self.set_of_stacks[index].pop()
            if len(self.set_of_stacks[index]) == 0:
                self.set_of_stacks.pop(index)
            return last_element
        else:
            self.set_of_stacks.pop(index)
            
    def printStacks(self):
        print(self.set_of_stacks)
```

Write a program to sort a stack such that the smallest items are on the top. You can use an additional temporary stack, but you may not copy the elements into any other data structure (such as an array). The stack supports the following operations: push, pop, peek, and isEmpty.

```python
class Node:
    def __init__(self,value):
        self.value = value

class Stack:

    def __init__(self):
        self.stack_list = []

    def push(self, value):
        '''Push element to the top of stack'''
        new_node = Node(value)
        self.stack_list.append(new_node)
        
    def stack_pop(self):
        if self.isEmpty() == True:
            print("No element in the Stack")
            return 
        else:
            return self.stack_list.pop()
            

    def peek(self): 
        '''Return top element of the stack'''
        
        return self.stack_list[-1]
    
    def isEmpty(self): 
        '''Return False if stack is not empty'''
        if len(self.stack_list) == 0:
            return True
        else:
            return False
    
    def printStack(self): 
        '''Print all elements of the stack'''
        for element in self.stack_list:
            print(element.value)
      
```

```python
def sortStack(stack):

    sortedStack = Stack()
   
    first_value = stack.stack_pop().value
    sortedStack.push(first_value)

    while stack.isEmpty() != True:

        if stack.peek().value < sortedStack.peek().value:
          
            higher_value = sortedStack.stack_pop().value
            lower_value = stack.stack_pop().value

            stack.push(higher_value)
            sortedStack.push(lower_value)
        
        if stack.peek().value > sortedStack.peek().value:
            higher_value = stack.stack_pop().value
            sortedStack.push(higher_value)
        

           
            
    

    return sortedStack
    
   
    
```

# 3.6 Animal Shelter: Correct but solution uses a only a dog queue and a cat queue and has timestamp for each animal 
An animal shelter, which holds only dogs and cats, operates on a strictly"first in, first out" basis. People must adopt either the "oldest" (based on arrival time) of all animals at the shelter, or they can select whether they would prefer a dog or a cat (and will receive the oldest animal of that type). They cannot select which specific animal they would like. Create the data structures to maintain this system and implement operations such as enqueue, dequeueAny, dequeueDog, and dequeueCat. You may use the built-in Linked List data structure.

```python
class Animal:
    def __init__(self,name,animal_type):
        self.animal_type = animal_type
        self.name = name

class AnimalShelterQueue:
    def __init__(self):
        self. all_queue = []
        self.dog_queue = []
        self.cat_queue = []
    
    def enqueue(self,name,animal_type):
        
        new_animal = Animal(name,animal_type)
        
        #add animal to all queue 
        self.all_queue.append(new_animal)
        
        #add animal to specific queue
        if new_animal.animal_type == 'dog':
            self.dog_queue.append(new_animal)
        if new_animal.animal_type == 'cat':
            self.cat_queue.append(new_animal)
            
    def dequeueAny(self):
        popped_animal = self.all_queue.pop(0)
        if popped_animal.animal_type == 'dog':
            self.dog_queue.pop(0)
        if popped_animal.animal_type == 'cat':
            self.cat_queue.pop(0)
        return popped_animal
        
    
    def dequeueDog(self):
        self.all_queue.pop(0)
        return self.dog_queue.pop(0)
    
    def dequeueCat(self):
        self.all_queue.pop(0)
        return self.cat_queue.pop(0)
        
        
            
        
            
        
    
```
