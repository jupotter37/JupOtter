```python
def groupAnagrams(strs):
    # convert to list so we can iterate over without consuming (map returns a consumable structure that is exhausted after iteration via enumerate)
    sortedstrs = list(map(lambda x: ''.join(sorted(x)), strs))
    matches = []
    visited = []

    for i,s in enumerate(sortedstrs):
        if i in visited:
            continue

        loopmatches = [strs[i]]

        for sorti, sortd in enumerate(sortedstrs):
            if s == sortd and i != sorti:
                visited.append(sorti)
                loopmatches.append(strs[sorti])
            
        matches.append(loopmatches)
    
    return matches

strs = ["eat","tea","tan","ate","nat","bat"]
matches = groupAnagrams(strs)
print(matches)
```

```python
# O(n) time and O(1) space
# Main idea is to get numbers before and after the current index in nums (use prefix product and postfix product and then multiply them together)

def productExceptSelf(nums):
    # to get O(n) time we can use two for loops (O(2n) == O(n))
    res = [1] * len(nums) # initialize results to length of the list passed in
    # we will use this results list to store prefixes (product before current i) on one pass and then reverse pass on nums to get the postfix (after current i) and multiply that by the prefixes
    
    prefix = 1 # initialize prefix (for first prefix product we use 1 since there is nothing before the first entry in nums)
    # get prefixes to the current index (the product of all numbers before the current index)
    for i in range(len(nums)):
        res[i] = prefix # assign the prefix product to the current index and save it
        prefix *= nums[i] # get product of prefix so far and mult by current number to get and store the prefix product for the next index
    # now get postfixes by iterating from the end of nums and multiply it by the saved prefix products in results list
    postfix = 1 # init to 1 since there is nothing after the last index, this will get updated as we walk backwards in nums
    for i in range(len(nums)-1,-1,-1):
        # we multiply the postfix products by starting at end and go backwards to get products after i stored
        res[i] *= postfix # multiply the stored prefix (product of nums before i) by the current postfix to get products for that index (excl. index)
        postfix *= nums[i]
    
    return res

print(productExceptSelf([1,2,3,4]))
```

```python
# Longest sequence of consecutive numbers
# main idea is to check if the element in a set (to remove duplicate vals) has a left neighbor that is one less than it. If no, it means that it is the start of a sequence and you can start checking for consecutive numbers after it

# O(n) time complexity (we visit each element at most 2 times 2n breaks down to n)
# memory complexity is O(n) since we create the set from n els in a list

def longestConsecutive(nums): 
        '''returns a number - the longest number of els that make a consecutive sequence in the list (the els can be in any order and contain duplicates.)'''
        if len(nums) == 0:
            return 0
        
        # convert list to a set to remove duplicate numbers
        numSet = set(nums)
        longest = 0 # initialize a running count of the longest sequence found

        for n in numSet:
            # check if the number has a left neighbor that is 1 less than it - if not, it means that it is a start of a new sequence:
            if (n-1) not in numSet:
                # track the length of the current sequence
                length = 0
                # increase additions to start incrementally by 1 and check for the number in the set until failed to get the length of the sequence:
                while (n+length) in numSet:
                    length += 1
                # update the longest running count if length is greater than current recorded longest
                longest = max(length,longest)
        
        return longest
```

```python
import re

# uses more memory
def isPalindrome(s: str) -> bool: 
    cleaned = re.sub(r'\W|_+',"",s)
    lowercase = cleaned.lower()
    reversed = lowercase[::-1]
    return reversed == lowercase

isPalindrome("ab_a")

# this solution uses O(1) memory - we use pointers at the beginning and end of string to check and match chars
def lowmemoryPalindrome(s):
    # create pointers on ends of string
    l = 0
    r = len(s) - 1 # last char pointer
    # go through string until pointers meet or pass each other
    while l < r:
        # skip chars that are not alphanumeric
        while l < r and not alphaNum(s[l]):
            # increment left pointer past all nonalpha chars:
            l += 1
        while r > l and not alphaNum(s[r]):
            r -= 1 # decrement right pointer to skip all nonalpha chars
        # compare chars and return false if no match
        if s[l].lower() != s[r].lower():
            return False
        # update pointers
        l += 1
        r -= 1
    # we've reached end of str and not returned early so all chars match,
    return True

# helper to check if alpha numeric - we use ord to get the ASCII number
def alphaNum(c):
    return (ord('A') <= ord(c) <= ord('Z') or
            ord('a') <= ord(c) <= ord('z') or 
            ord('0') <= ord(c) <= ord('9')) # check if c is in between or equal to alpha numeric chars

lowmemoryPalindrome("ab_a")
```

```python
### Two Sum (unsorted)

## The main idea is to start with a empty hash map and start saving values seen as you walk as {value: index}
## This allows for O(n) time and O(n) memory space

def twoSum(nums,target):
    # create hash map empty initially
    prevMap = {}

    for i, n in enumerate(nums):
        diff = target - n
        if diff in prevMap:
            return [prevMap[diff], i] # return the inidices
        prevMap[n] = i

twoSum([2,7,11,23], 9)
```

```python
### Two Sum II (sorted)

# find two nums that add up to the target and return their indices (1 based)
# Brute force would be exponential time (O(n^2))
# using two left and right pointers is O(n) linear time
# The idea is to subtract ends of the list and increase the left pointer if the result is too small, and decrement the right pointer if the sum is too large
  # this allows us to iterate through the array 1 time and not use extra memory
  # takes advantage that the numbers are sorted in ascending order

def twoSum(numbers, target):
    l = 0
    r = len(numbers)-1

    while l < r:
        curSum = numbers[r] + numbers[l]
        if curSum > target:
            r -= 1 # reduce right side to reduce sum
        elif curSum < target:
            l += 1 # add more
        else:
            return [l+1,r+1]
    


        
result = twoSum([2,3,4],6)
result2 = twoSum([0,0,3,4],0)
result3 = twoSum([-1,-1,-1,-1,1,1],2)
print(result)
print(result2)
print(result3)
```

```python
### Three Sum

## Main idea is to sort the list, then use a updated hashmap as you walk to eliminate computing duplicate combinations and then use Two Sum II solution for the remaining 2 variables of the triplicate
## (as in two sum II we then use two pointers at each end of the sorted list to avoid extra mem with a hashmap)
# Time complexity uses two loops so is O(n^2), space complexity is O(n) due to sorting if it takes memory

def threeSum(nums):
    # sort the list first
    res = []
    nums.sort()

    for i, a in enumerate(nums):
        # skip duplicates (they will be next to each other in the sorted list):
        if i > 0 and a == nums[i-1]:
            continue # skip

        # use two sum II approach to solve for remaining two vars
        l = i + 1
        r = len(nums)-1

        while l < r: # l and r pointer cannot cross or be equal
            threeSum = a + nums[l] + nums[r]
            # if combo is greater than zero we need to decrease sum, so decrement right pointer
            if threeSum > 0:
                r -= 1
            elif threeSum < 0:
                l += 1 # increment left to make sum bigger if too small
            else:
                # collect result if equals zero (l and r did not cross and we found a solution)
                res.append([a,nums[l],nums[r]])
                # now move pointer on b value to get setup for the next a value
                # we only shift the left pointer (the other pointer will be updated by the above cases)
                l += 1
                # we don't want the same sum so we use a loop. If we find the same value, we need to keep shifting to skip duplicate a values
                while nums[l] == nums[l-1] and l < r: # also guard against l and r pointers crossing
                    l += 1
        
        return res




```

```python
# brute force O(n^2) - water capacity problem
def maxArea(height):
        maxcapacity = 0
        for i in range(len(height)):
            for j in range(len(height)-1,-1,-1):
              if i < j:
                  span = j - i
                  h = min(height[i], height[j])
                  capacity = span * h
                  if capacity > maxcapacity:
                      maxcapacity = capacity
        
        return maxcapacity


# O(n) linear solution
# Keep a left and right pointer, move the one over that has a lesser height (to retain the bar with the greatest height)
def maxArea(height):
    maxcapacity = 0
    # left and right pointers
    l = 0
    r = len(height) - 1

    # keep iterating until left crosses right:
    while l < r:
        span = r - l
        area = span * min(height[l],height[r])

        maxcapacity = max(area, maxcapacity)

        if height[l] < height[r]:
            l += 1
            continue
        if height[r] < height [l]:
            r -= 1 # decrement right pointer if it's the lesser
            continue
        # if they are equal then we move any pointer over
        l += 1

    return maxcapacity
        

maxArea([1,8,6,2,5,4,8,3,7]) # 49
```

```python
# Trapped rain water

# for O(1) memory, O(n) runtime solution, keep left and right pointers
# keep track of the maximum heights on the right or left
# The bottleneck for water trapped is the minimum of the max heights (water overflows if beyond this height)
# We shift the left or right pointer if they are the minimum of the two max heights
# The pattern focuses on the lesser of left or right heights and determines if water can be held adjacent to it (to the right of left, or to the left of the right block)

def trap(height):
    # edge if height empty
    if not height: return 0

    l = 0
    r = len(height)-1
    trapped = 0

    maxLeft = height[0] # on the boundary at beginning there is no left height container
    maxRight = height[-1] # boundary: no right container height to the right of the edge

    while l < r:
        # shift the pointer that is the current minimum of the max heights:
        if maxLeft < maxRight:
            l += 1            # update the max height if needed
            maxLeft = max(maxLeft, height[l])
            # calculate trapped water
            # If leftMax is less than (or equal to) height[l], we set leftMax equal to height[l] and then subtract height[l] from it, which would equal zero.
            # On the other hand, if leftMax is greater than height[l], we keep leftMax as is and subtract height[l] from it, which would be a positive value. 
            trapped += maxLeft - height[l] # this will never be negative since we update the left max before doing this computation
        else:
            r -= 1
            maxRight = max(maxRight, height[r])
            trapped += maxRight - height[r]

    return trapped

trap([4,2,0,3,2,5]) # 9
```

```python
def characterReplacement(s: str, k: int) -> int:
        counts = {}
        mostfrequentchar = 0
        maxcount = 0 # using this gives us O(n) over O(26n) since we don't have to scan counts (containing up to 26 letters) to get the maximum below.

        l = 0
        r = 0
        # use a sliding window with left and right pointers. increment the left if the window has least frequent chars that cannot be decremented less than or equal to the limit of k times.
        # find the char with the most counts and replace other elements k times with that char
        for r in range(len(s)):
            char = s[r]
            counts[char] = 1 + counts.get(char, 0) # increment recorded count for the char
            mostfrequentchar = max(mostfrequentchar,counts[s[r]])

            windowsize = r - l + 1
                        
            if (windowsize - mostfrequentchar) <= k:
                maxcount = max(maxcount,windowsize)
            else:
                counts[s[l]] -= 1
                l += 1
        return maxcount

print(characterReplacement("AABABBA",1)) # 4
```

```python
## Minimum Window Substring - find shortest window substring in s that contains all chars in t
def minWindow(s, t):
    # handle edge case:
    if t == "": return ""

    # make two maps of the counts of chars in t string and the count of those chars in our current window
    countT = {}
    window = {}

    # initialize the counts of our t string (this does not change after this)
    for c in t:
        # for every char in t add a count of 1 to it in our map
        countT[c] = 1 + countT.get(c, 0)

    # keep track of what counts of the chars we have in our window, and what counts we need (compare to countT)
    # we only need to track the chars in the t string since we only care about those in the window
    # the main idea is to compare the condition/counts of the characters in the maps and keep track of how many conditions (countT == countwindow) are satisfied
          # when we have all conditions satisfied (one for each distinct char in t), then we have a valid window
    have = 0 # how many conditions satisfied - the character count in window is the same as in countT is 1 have
    need = len(countT) # gives us how many unique chars are in t   

    res = [-1,-1] # default to -1s - will be a left and right pointer later
    resLen = float("infinity") # default val since we are looking for minimum length of substring, used to check length of windows to find shortest

    l = 0 # initialize left pointer

    # now start iterating through every char in s, r is for the right pointer/boundary of our window:
    for r in range(len(s)):
        c = s[r] # get the character at current right pointer
        # update the current window with the current char/count we've reached:
        window[c] = 1 + window.get(c, 0)

        # we only care about chars in t, check if the counts for the chars are equal in window and t
        if c in countT and window[c] == countT[c]:
            # since we satisfy the condition, we update that we have fulfilled one of the requirements for our window and t character/count
            have += 1 # note that this can hit +1 as soon as condition satisfied, future matching same chars in s can be added to window, but we don't change this then since we've already met the requirement
        
        # loop if our have is equal to the need conditions (char counts are the same for all t chars and window)
        # if we have all the matches need, we want to slide our left pointer over until we no longer have all the matches because this will give us the minimum size of the window that does have all the matches we need. 
        # When the window no longer has all the matches we want we continue iterating the right pointer
        while have == need:
            windowSize = r - l + 1 
            if windowSize < resLen: # executes at least once since we default to infinity above
                # if we find smaller window that satisfies all conditions, update the result
                res = [l, r] # as soon as we have all conditions, set the boundaries for current window
                resLen = windowSize
            # while all conditions met we want to keep shrinking the window to smallest possible satisfying all conditions
            # pop from left of the window
            window[s[l]] -= 1 # decrement the left most char in window map count by 1
            # it is now possible all conditions are not met anymore, so check condition:
            if s[l] in countT and window[s[l]] < countT[s[l]]:
                # if we just made out char count less than needed from decrementing, then let's remove a have condition satisfied (we have one less condition fullfilled now)
                have -= 1
            # finally shift the left pointer of window over since we removed the char from our window
            l += 1 # note we don't update the res left pointer because if we remove a needed condition, we don't want to update our result window
            # now the loop will check if have equals needed conditions and re-execute as needed
    
    # now return the result
    l, r = res # extract pointers of shortest window
    return s[l:r+1] if resLen != float("infinity") else "" # return substring, add 1 for off by one (includes last element represented by right ptr since slice is exclusive) on right side and only return if result has been changed from default

        
# this demonstrates need for while loop:
minWindow('bbaa','aba') # should return 'baa', without while loop returns bbaa
        





        
```

```python
import collections

# O(n) time and O(n) memory due to using a queue data structure

## Max Sliding Window
# use a monotonically decreasing queue approach (store nums as scanned in a deque in decreasing order, pop the lesser ones on the right off) - left most val is the highest, smaller values are added to the right of it.
# using the deque/queue you can add and remove elements in O(1) time as you scan resulting in an overall op of O(n)

def maxSlidingWindow(nums, k):
    output = [] # store maxes to return
    # initialize left and right pointers to 0
    l = 0
    r = 0
    # create a queue to store the indices of numbers in window - max on the right
    q = collections.deque()

    # loop while right pointer is still in bounds
    while r < len(nums):
        # ensure no smaller values exist in the queue
        # remove smaller values from the queue
        # furthest right in q is the index of the highest number? or is that backwards?
        current = nums[r]
        while q and nums[q[-1]] < current: # the last item added is less than current, we want to get rid of it and keep popping as long as other els are less than current
            q.pop() # pops from right of queue (the number that is less than what we found currently in the window, we remove since we're interested in maximum number)
        # now we can add the new val to the queue
        q.append(r) # curent is less than what's in q, we add it to queue (assembling the nums in descending order)

        # as sliding window updates, remove the out of bounds left from the queue:
        if l > q[0]: # left bound of window index becomes more than the left most val(index) stored in the queue
            q.popleft() # pop the out of bound index stored from left side of queue
        
        # edge case: since we start at 0 indices for l and r, we have to make sure the window is at least size k
        if (r+1) >= k:
            # append the maximum val to the output
            output.append(nums[q[0]]) # the left most val in q is the index of the maximum for the window
            # left is only incremented when window is at least size k
            l += 1

        r += 1
    return output

maxSlidingWindow([1],1)
```

```python
## VALID PARENTHESES

# ((())) or ([]) is valid - must be in order and everything closed.

def isValid(s):
    stack = []
    # create a hashmap of closing to opening parens - if top of stack matches with an opener, then we can pop off stack until empty - that means s is valid
    closedToOpenMap = {
        ']': '[',
        ')': '(',
        '}': '{'      
    }

    # trick is to determine if char is a closing paren or not
    # if it's an opening paren we can add always to stack, if closing then the top of stack must match the opening for it

    for c in s:
        # check if it's a closing paren
        if c in closedToOpenMap:
            # make sure stack is not empty and top of stack is matching opener
            if stack and stack[-1] == closedToOpenMap[c]:
                stack.pop()
            else:
                # if no match then string is invalid because we can't start with closer
                return False
        else:
            # if we have an opener we can add as many as we want to the stack
            stack.append(c)

    return len(stack) == 0 # if emtpy everything has been matched and closed
```

```python
def generateParenthesis(n: int):
    res = []
    i = 0
    
    def backtrack(open_n, closed_n, path):
        nonlocal i
        i += 1
        current = f'bt {i} - {open_n=}, {closed_n=}, {path=}'
        print(f'entering {current}')
        if open_n == closed_n == n:
            print('\n')
            print(f'base case hit, appending res and exiting {current}\n\n')
            res.append(path)
            return
        
        if open_n < n:
            # recurse and increment open count
            print(f'openn < n, entering {i+1}')
            backtrack(open_n + 1, closed_n, path + "(")

        if closed_n < open_n:
            print(f'closedn < openn, entering {i+1}')
            # recurse and increment closed count
            backtrack(open_n, closed_n + 1, path + ")")
        print(f'exiting {current}')
            
    backtrack(0, 0, "")

    return res

generateParenthesis(2)
```

```python
# Ideal Numbers in a Range
  # how many numbers in a range of a low and a high (inclusive) are the result of 3^x x 5^y
# adapted from this solution: https://www.geeksforgeeks.org/count-numbers-from-range-whose-prime-factors-are-only-2-and-3/


def findidealnums(low, high) :
    count = 0
 
    for i in range(low, high + 1) :
        if i % 2 == 0: # skip even nums
            continue

        num = i
 
        # While num is divisible by 3, 
        # divide it by 3
        while (num % 3 == 0) :
            num //= 3
 
        # While num is divisible by 5, 
        # divide it by 5
            # note that we continue this operation after the number has been reduced above with the factor of 3
        while (num % 5 == 0) :
            num //= 5
 
        # If num got reduced to 1 then it has
        # only 3 and 5 as prime factors, also covers edge case of 1,1
        if (num == 1) :
            count += 1
 
    return count

test1 = findidealnums(200,405) # 4
test2 = findidealnums(1,1) # 1
test3 = findidealnums(1,10) # 4

print(f'{test1=}, {test2=}, {test3=}')
```

```python
# Car Fleet problem: cars with positions and speeds - how many become fleets by end reaching the target
# if car catches up to another cars go the speed of the car in front (the slower speed)
# https://leetcode.com/problems/car-fleet/

def carFleet(self, target: int, position, speed) -> int:
    # O(nlogn) - go through cars once plus sort them.
    # O(n) memory 
    # Main idea: sort cars by position and compare distance to end divided by speed to get the time they arrive at target. 
        # If the previous behind car gets there earlier or same time, then it becomes part of a fleet and collides with car in front at some point
        # Remove behind cars since the new speed is the car in front that got collided with (eliminates tracking new speed etc.)
        # By focusing on comparing time and end destination we make a calculation per pair once and only have to scan once

    # Algo:
    # sort cars by position (ascending order)
    # start from the right, go through cars in reverse order because we don't know what the car to the right's speed will be throughout the process (we always know the right most car's speed)
    # Start with empty stack, add last car to it, then next one back and start comparing dist/time to target.
    # pop the behind car in the pair (top of the stack) when it reaches target before or same time as car in front of it

    # create list of pairs of position and speed for the cars
    pair = [[p,s] for p, s in zip(position, speed)]
    # init stack
    stack = []

    # sort the list by position in ascending order
        # -sorted will sort by first item in pairs (position)
    for p, s in sorted(pair)[::-1]: # reverse order to start with car in last position
        # use decimal division to find the time the car will be at the target and append to stack:
        stack.append((target - p) / s) # dist from target divided by speed to get units of time when it arrives there
        
        # we don't want to compare until we have at least two cars - otherwise do nothing
        if len(stack) >= 2 and stack[-1] <= stack[-2]: # use if not while - we check each collision right to left and don't need a while
            # if there is a collision (the top of stack is car behind the other one) then pop it from stack
            stack.pop() # this leaves the car in front previously on the stack in the stack
    
    # how many collisions indicate how many fleets at the end
    return len(stack)
```

```python
### BINARY SEARCH

# Runs in O(logn) time
  # log_2(n) => yields how many times we can divide n by 2 (instead off and much less than how many n's there are)

# need a sorted list
# use two pointers, find midpoint (// 2), stop if midpoint == target, otherwise search half of list based on whether greater or lesser than midpoint

def search(nums, target: int) -> int:
    l = 0
    r = len(nums) - 1

    # if left crosses right it means we did not find the result since it shifted outside of list
    # if left and right are equal it means we have not looked at that value yet
    while l <= r:
        # find mid point
        # side note: if avoiding buffer overflow in other languages with large lists, you can do: l + ((r-l) // 2) - this keeps numbers under the r value instead of adding to it and still gets the midpoint
        mid = (l + r) // 2 # int division floors result - will be one to the left for even numbered list lengths
        # if the midpoint is larger than the target, move the right pointer (end) to shrink the list by half
        if nums[mid] > target:
            r = mid - 1
        elif nums[mid] < target:
            # target is on right half, move left pointer up after mid
            l = mid + 1
        else:
            # if not greater or less than, we found the target
            return mid # return the index
    
    # if not found, return -1
    return -1

search([-1,0,3,5,9,12],9) # 4


### REcursive:
def recsearch(nums, target):
    if len(nums) == 0: # base case - could also try when l <= r
        return -1
    mid = (len(nums)) // 2

    if nums[mid] == target:
        return mid
    elif nums[mid] > target:
        return search(nums[:mid], target)
    else:
        result = search(nums[mid+1:], target)
        startIdx = mid + 1
        if result != -1:
            return startIdx + result
        else:
            return -1
        

print(recsearch([-1,-0,3,5,9,12],9)) # 4
```

```python
# 2-d matrix binary search

def twoDMatrix(matrix,target):
    ROW, COL = len(matrix), len(matrix[0])
    left = 0
    right = ROW * COL - 1

    while left <= right:
        mid = (left + right) // 2
        row = mid // COL
        col = mid % COL

        if matrix[row][col] > target:
            right = mid - 1
        elif matrix[row][col] < target:
            left = mid + 1
        else:
            return True
    return False



# twoDMatrix([[1,3,5,7],[10,11,16,20],[23,30,34,60]],3)
twoDMatrix([[1,2],[3,4]],3)
```

```python
## KOKO's BANANAS

# main idea is to take highest rate koko needs to eat and then work backwards towards 1 to find if there is a lower k to satisfy time limit
      # since the constraint of piles <= hours, we know that if we choose the highest pile count as the rate/hour we can eat all piles in time.
    # Take the highest numbered pile and that would be the max rate koko can eat per hour (but we want to find if there is a lower rate to satisfy time limit)
    # Use binary search on the range from 1 to the highest pile [1...11 etc.] (instead of on the actual piles list) 
      # if Koko can eat all in time limit from the midpoint of that range, then search to the left, if not, search to the right for a higher rate that satisifies time
    # Gives us O(log(max(p)) * p) runtime

import math

def kokosBananas(piles,h):
    maxP = max(piles)
    l = 1
    r = maxP
    minK = maxP # initialize min not to zero, but to the max we could possibly be and acheive time limit


    while l <= r: # cannot cross - then we've finished binary slicing everything
        k = (l+r) // 2 # set k (rate of eating bananas per hour) to the midpoint
        totalHours = 0

        for pile in piles:
            totalHours += math.ceil(pile / k) # round up since once pile is empty koko cannot eat from another pile for the rest of the hour
            if totalHours > h:
                break
        
        if totalHours > h:
            l = k+1 # we need to search right, the rate was not fast enough so we need a faster/larger rate
        else:
            minK = min(minK, k) # rate satisfies time, so record minK and search left to try to find a smaller one
            r = k-1

    return minK

print(kokosBananas([3,6,7,11],8)) # 4 is the lowest k
```

```python
# compound interest - using recursion

import locale
print(locale.getdefaultlocale()[0]) # 'en-US

locale.setlocale( locale.LC_ALL, '' ) # default locale must be set

def currency(n):
    return locale.currency(n, grouping=True )

def compound(init_amount, rate, years):
    year_count = 0
    amount = init_amount

    # uses closure to encapsulate count and total
    def calculate():
        nonlocal year_count 
        nonlocal amount

        if year_count == years:
            return amount

        before = amount
        amount = amount + (amount * rate)

        year_count += 1
        print(f'Interest earned after year {year_count}: {currency(amount - before)}')

        return calculate()
    
    return calculate

final = compound(10000,0.0435,10)()
print(f'Total after compounding: {currency(final)}')
```

```python
# Min in rotating numbers list
# use binary search
# O(logn)

# the key is to look at the ends of the arr in binary search (i.e. the l and r pointers) to determine where we are in the rotation (if end is less or greater than the midpoint reveals the answer)
def findMin(nums) -> int:
    left , right = 0 ,len(nums) - 1 

    # note that we don't do <= here in the while condition. when left == right that is the pivot (we've eleminated other possibilities, so we just return nums[left])
    while left  <  right:
        mid = (left + right) // 2
        if(nums[mid+1] < nums[mid]): # if next num to right of mid is less, thats the pivot/rotate point/minval
            return nums[mid+1]
        
        # Check the end of the right half - if mid is greater it means the pivot must be to the right
        if nums[mid] > nums[right]:
            left = mid + 1
            
        # left has the  min 
        else:
            right = mid # note not mid-1 - the mid could be the actual pivot we're looking for so include it.
            
    return nums[left]

findMin([5,1,2,3,4]) # 1
findMin([4,5,1,2,3]) # 1
```

```python
# Find target in a rotating list

# O(logn)
# use binary search - key is to check target against mid AND the ends of the array (left end or right end) to determine which way to search

def search(nums, target):
    # init left and right pointers
    l = 0
    r = len(nums)-1

    # if we have one element ([1]), then left and right will be equal and we still need to check the value
    while l <= r:
        mid = (l+r) // 2
        # mid could be target, so check and return
        if target == nums[mid]:
            return mid
        
        # check if we are in the left sorted portion
        if nums[l] <= nums[mid]: # possible that mid and left val could be the same
            # if target is greater than mid val, you need to search right
            if target > nums[mid]:
                l = mid+1
            # check if target is less than middle, but also less than the left most value
            elif target < nums[l]:
                # need to search the right
                l = mid+1 # could combine this with an or above.
            else:
                # the target is less than the middle, but greater than the left end - search left
                r = mid - 1
        else:
            # we're in the right sorted portion
            if target < nums[mid]: # search left half
                r = mid-1
            elif target > nums[r]: # target is greater than middle and greater than the right most val
                # need to go left as well
                r = mid - 1
            else: # target is greater than mid and less than right most val
                # just search right
                l = mid + 1
            
    return -1
```

```python
### REVERSE LINKED LIST

# O(n) time and O(1) memory (we only store pointers, no data structures etc.)

# use two pointers - save the current node to reference to go to the next node before breaking and resetting it's next
# reset the next of every node to point to the previous node in the original list

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def reverseList(head):
    # use two pointers
    curr = head
    prev = None # initialize to None as this will be the next val for the last node in the reversed list

    while curr:
        # while the curr is not null, reverse the pointers

        # save the current next in a temp var so when we update curr.next we don't lose this
        nxt = curr.next # save the original next of the current (we use this to move the pointer to it for next iteration)
        curr.next = prev # will be None on first iter. points current next to previous modified node (from last iteration)
        prev = curr # update prev pointer (has next set to the prev node)
        curr = nxt # move current pointer to the next node in the list. (will be None on last iter and stop loop because nxt will be None on the last Node in list)
    
    # return after the loop - the result will be stored in prev
    #  the two pointers march through the nodes such that prev starts with None and head starts with the first node, 
    # so when the loop finishes, prev points to the last node and head points to None, so we definitely need to return prev 
    # as the new head (in the last iteration prev gets updated to the last node and the head gets updated as None)

    return prev # covers head is null case as well due to while condition

head = ListNode(1,ListNode(2))
vars(head.next)
```

```python
# Splice/merge two linked lists:

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

def mergeTwoLists(list1, list2):
    # if equal add both and advance both pointers
    # while the lesser of two is less than the greater, advance that one and append

    # common technique is to insert a dummy node at the beginning (any value) to avoid edge case of inserting into an empty list
    dummy = ListNode()
    tail = dummy # holding the remaining list after dummy insertion

    # if both lists are non-empty we can compare values
    while list1 and list2:
        if list1.val < list2.val:
            # insert list1
            tail.next = list1
            list1 = list1.next # update pointer
        else: # list2 < list1
            tail.next = list2
            list2 = list2.next
        # update tail pointer regardless
        tail = tail.next

    # if one list is empty and the other isn't now, find nonempty list and insert the rest of it to the end
    if list1:
        tail.next = list1
    elif list2:
        tail.next = list2
    
    return dummy.next # head starts after the inserted dummy node.
```

```python
## Reorder Linked List: 1,2,3,4 > 1,4,2,3
    #  can be achieved by splitting the linked list into 2 halves, reverse the second half then merge it in the first half.

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

def reorderList(head) -> None:
    """
    Do not return anything, modify head in-place instead.
    """
    # Pattern: use a slow and fast pointer (to find the middle divide of the linked list)
    # Divide the list and reverse the second half of the list

    
    # O(2n) > O(n) - we iterate through the list twice

    slow = head
    fast = head.next

    # while the fast pointer has not reached the end of the list
    # covers if there is only one item in linked list as well if head.next is null
    while fast and fast.next:
        # increment slow by 1 and fast by 2 - this will stop when slow is at the middle of the list
        slow = slow.next # will be midpoint when loop is done
        fast = fast.next.next # increment double the distance of slow by 2
    
    # get the second half of the list (mid point will be considered the node after where we stopped)
    second = slow.next
    # split off/break off the first half of the list
    slow.next = None
    prev = None # initialize the previous node to null for use in reversal below
    # reverse second half of list
    while second:
        # store original next
        tmp = second.next
        # reverse connection to the previous node (will be null initially so we're setting the first el in the second half to null since it will be the end node after reversal)
        second.next = prev
        # set the previous in the path to be the current node being operated on
        prev = second
        second = tmp # set to next node in list
    
    # merge two halves of list
    second = prev # will be the start/head of the second half (reversed) and second will be null at this point
    first = head # the original head of the list

    # until one of the pointers is null
    # the second half could be shorter depending on the mid (if odd numbered list etc)
    while second:
        # save nodes in each list
        tmp1 = first.next
        tmp2 = second.next

        first.next = second
        second.next = tmp1 # inserting in between first and first.next
        # shift pointers
        first = tmp1
        second = tmp2
```

```python
# Lowest common ancestor tree problem

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

# "lowest" means lowest in the tree, i.e. furthest down and still common to both nodes.
# O(logn) time complexity - we visit node in level so the height of the tree is the determinate factor of input
# memory complexity is O(1) - we don't need any data structures etc to store anything
# main idea is to leverage the BST ordering and go left or right depending on if p and q are greater than or less than a current node
# if both are greater than current node, then lowest ancestor must be to the right, and vice versa if both less than
  # else case: if one is either equal to current node or less than or greater than, then the current node has to be the ancestor
  # this means there is a split and the nodes are either equal to current or on different sides of the tree, so there can't be a further common ancestor at this point.
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # start at the root node
        cur = root
        # continue until you find the result
        while cur: # while not null, gauaranteed result in restrictions
            # leverage BST ordering to decide which way to search - all children on left will be less than current node, all children on right will be greater than current node
            # if both greater, then you search right-
            if p.val > cur.val and q.val > cur.val:
                # update the current pointer to be the right side of tree
                cur = cur.right
            # note that you need the elif and else conditions so we don't return cur everytime or under wrong circumstance
            elif p.val < cur.val and q.val < cur.val:
                # update to go to the left
                cur = cur.left
            else:
                # if a split occurs or we find p or q:
                # we have found the result so return the current node
                return cur
            
            
```

```python
# Record path left to right of a tree

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
import collections

# O(n) time complexity
# Use BFS to add levels to a result list one level at a time
  # we separate level population by iterating over it per level and popping each iter. So it is cleared at the end of each level of the for loop, but pre-populated with all children for the next loop
    # top control of the loop is based on whether the queue is empty or not.
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        # list to hold final path levels
        res = []

        # use a queue for the BFS
        q = collections.deque()
        # initialize the queue with the root node
        q.append(root)

        # when queue is empty we know we've gone through all the levels
        while q:
            # queue will hold the current level so we want to iterate over that number of items
            qLen = len(q)
            level = [] # holds the current level values to add to our result
            # loop over current level
            for i in range(qLen):
                # get top of queue
                node = q.popleft()
                # we can have one or more nodes on the level be empty, we don't want to add nulls to our results
                if node: # prevent null being inserted into level list
                    level.append(node.val) # add current node to our level one by one in the current loop
                    # we add to the end of the queue all children of the current level which does not count this time, but next round, we'll one by one add those to their own level in the next loop
                    q.append(node.left) # adding child to queue for next loop iteration/level
                    q.append(node.right)
            # We need this to prevent emtpy intial level list from being added.
            if level: # only append if there are nodes in the level list
                res.append(level)
        # after queue is empty we have traversed all levels
        return res
```

```python
# Word Search (Graph with backtracking and recursion)

class Solution:
    def exist(self, board, word) -> bool:
        # use recursive (dfs) backtracking pattern
        # edge case: row could be empty - we will have at least one row 
        # brute force: O(n * m * 4^len(word))
        
        ROWS = len(board)
        COLS = len(board[0])
        # record path to make sure you don't visit the same cell again
        path = set()

        # start backtracking pattern
        # create dfs inner func to have access to variables in parent func (board etc)
        # pass in position (r,c) and current char in target word we're looking for - i
        def dfs(r, c, i):
            # base case - we found the word:
            if i == len(word):
                return True
            
            # we're out of bounds, char not in word or we already visited the position:
            out_of_bounds = r < 0 or c < 0
            over_bounds = r >= ROWS or c >= COLS
            already_seen = (r,c) in path

            if (out_of_bounds 
                or over_bounds 
                or word[i] != board[r][c] 
                or already_seen):
                return False
            
            # we found a char in the word
            # add to our path
            path.add((r,c))
            # start recursing - move pos and look for the next char in word
            # search all four directions - one has to return True when word is found
            res = (dfs(r+1, c, i+1) or 
                   dfs(r-1,c,i+1) or
                   dfs(r,c+1,i+1) or
                   dfs(r,c-1,i+1))
            
            # remove from path?
            path.remove((r,c))

            return res
        
        # brute force go through every position in grid
        for r in range(ROWS):
            for c in range(COLS):
                beginning_of_word = 0
                # if dfs ever returns True immediately return True - word found
                if dfs(r,c,beginning_of_word): return True
        
        # word was never found return false
        return False


## Without comments:

class Solution:
    def exist(self, board, word) -> bool:
        ROWS = len(board)
        COLS = len(board[0])
        path = set()

        def dfs(r, c, i):
            if i == len(word):
                return True

            out_of_bounds = r < 0 or c < 0
            over_bounds = r >= ROWS or c >= COLS
            already_seen = (r,c) in path

            if (out_of_bounds 
                or over_bounds 
                or word[i] != board[r][c] 
                or already_seen):
                return False

            path.add((r,c))

            res = (dfs(r+1, c, i+1) or 
                   dfs(r-1,c,i+1) or
                   dfs(r,c+1,i+1) or
                   dfs(r,c-1,i+1))

            path.remove((r,c))

            return res
        
        for r in range(ROWS):
            for c in range(COLS):
                beginning_of_word = 0
                if dfs(r,c,beginning_of_word): return True
        return False
```

```python
# Skyline problem
# https://leetcode.com/problems/the-skyline-problem/
# python solution: https://codereview.stackexchange.com/questions/221178/python-program-to-solve-the-skyline-problem

# Main idea is to look at left points created which are found when the highest building seen so far changes
# CAn use a priority queue (heap) to get O(1) of the tallest building and O(log n) time to insert/remove
  # when new building seen, push it on to active list
  # update skyline with the current x value and whatever the new tallest height is
  # buildings can have the same left corner point
# Right corners of buildings:
  # a second loop to get the right endpoints of buildings

# Try with [[2,5,3],[4,5,4],[6,7,3]]
  # result should be: [[2,3],[4,4],[5,0],[6,3],[7,0]]

#  Intuition: we are just maintaing the heights and if any new heights comes then we are adding it to ans and making sure that same heights change should not get added to ans.

# related algorithm: Sweep line - https://codeforces.com/blog/entry/20377
  # vertical line that sweeps from left to right and records values along the way (i.e. to track maximum height as the sweep progresses for example)

# not optimal solution? try using a heap
class Solution(object):
    # key is finding the upper left most points of the buildings
    def getSkyline(self, buildings):
        """
        :type buildings: List[List[int]]
        :rtype: List[List[int]]
        """
        if not buildings:
            return []
        if len(buildings) == 1:
            # second basecase - only one building
            # [left-x-position,height]
            left_endpoint = [buildings[0][0], buildings[0][2]] # the top left corner of the building (left x and height)
            right_endpoint = [buildings[0][1], 0] # the right end of building with y = 0 as specified in rules
            return [left_endpoint, right_endpoint]

        mid = len(buildings) // 2
        # recurse left and right
        # divide list - eventually hit base case and get the left upper corner and lower right corner of each building
        # left and right halves of the list of buidings
        left = self.getSkyline(buildings[:mid]) # recurse left of mid
        right = self.getSkyline(buildings[mid:]) # recurse right (include mid)

        # note: we don't hit merge if buildings list has one building (returns from base case)
        return self.merge(left, right)

    def merge(self, left, right):
        # track heights of skyline points
        h1, h2, hcurrent = 0, 0, 0
        i, j = 0, 0
        result = []

        # stop after getting to the end of one of the lists
        while i < len(left) and j < len(right):
            # left x edges of current bldg in left/right list
            x0 = left[i][0] # left x position of a building in the left list
            x1 = right[j][0] # left x-position of a building in the right list
            
          
            if x0 <= x1: # left-x of skyline pts in left list is less than the left-x of the right skyline point (left edge starts before)
                h1 = left[i][1] # height of current left skyline point (is 0 for the last point)
                i += 1 # move pointer to next bldg in left list

            # the right sklyine point for the left current i is at or more than the left x-position in the right list i
            if x1 <= x0: # point in the right list has left edge that starts before or at the skyline point x-pos in left list
                h2 = right[j][1] # record height of skyline point in right list
                j += 1
            if max(h1, h2) != hcurrent:
                hcurrent = max(h1, h2) # set curr height to tallest point hgt in current iter to get point in final skyline
                # add result point - the left most x-position with the current max height recorded
                result.append([min(x0, x1), hcurrent])

 
        # one of these or both of left or right will be out of bounds here, so we could be adding the part from one not iterated or nothing.
        # note the order of right extension before left - does this matter and why?
        result.extend(right[j:]) # doesn't need any extra checks to avoid producing two consecutive points at the same height
        result.extend(left[i:])
        return result
```

```python
# Number of Islands
def numIslands(grid) -> int:
    num_islands = 0

    directions = [(1,0),(-1,0),(0,1),(0,-1)]
    # here we will modify the input, if that is not allowed you need to create a set data structure to store visited

    def set_island_zeros(grid, r, c):
        # need the 0 in the range check to account for -1s in the directions - this check accounts for left and right bounds
        if ((0 <= r < len(grid)) and # not out of bounds row wise
            (0 <= c < len(grid[0])) and # not out of bounds col wise
            grid[r][c] == "1"): # don't recheck 0s since they are also "visited"
            # if it's a "1" check down up left and right (in recursive loop), set to "0" if another "1" encountered and search down up left right from there as well.
            # if "0" do nothing, no search from that point, just return out of recursive stack position

            # set all 1s to 0s so we don't recount or get infinite loop
            grid[r][c] = "0"
            # check directions
            for row_inc, col_inc in directions:
                set_island_zeros(grid, r + row_inc, c + col_inc)


    for row in range(len(grid)):
        for col in range(len(grid[0])):
            # the recursion will set the "1" to "0" as it finds adjacent land, so we won't search over area convered if the position is "0" (has been set to that)
            if grid[row][col] == "1":
                num_islands += 1

                set_island_zeros(grid, row, col)
    
    return num_islands
```

```python
### Trie implementation

# need to implement a Trie node
class TrieNode:
    def __init__(self):
        self.children = {} # can use a hashmap instead of array might be easier
        # children["a"] = TrieNode() node for a
        self.endOfWord = False # need to mark if char is end of a word

class Trie:
    def __init__(self):
        self.root = TrieNode() # just create a root node that is empty
        

    def insert(self, word: str) -> None:
        cur = self.root # used as a pointer until we get to last char in word to mark it

        # iterate through every char
        for c in word:
            if c not in cur.children: # not in the hashmap yet
                # create a node for this char
                cur.children[c] = TrieNode() # insert an empty node for the key of char
            # if char does exist, set cur to the child in children
            cur = cur.children[c] # update the pointer so we get to the end of the word and can mark it
        cur.endOfWord = True # mark cur set to last char as end of word

    def search(self, word: str) -> bool:
        cur = self.root # initialize a pointer to root of tree
        # go char by char and check if the word is in trie
        for c in word:
            if c not in cur.children:
                return False
            cur = cur.children[c]
        # cur will be the last character if chars in the trie
        return cur.endOfWord # we found the end marker meaning word is in trie

    def startsWith(self, prefix: str) -> bool:
        cur = self.root

        for c in prefix:
            if c not in cur.children:
                return False
            cur = cur.children[c]

        return True
```
