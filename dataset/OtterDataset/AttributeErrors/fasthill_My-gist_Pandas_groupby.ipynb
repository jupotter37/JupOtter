```python
total = country_grp['LanguageWorkedWith'].count()
selected = country_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
python_df = pd.concat([total, selected], axis=1)
python_df.columns = ['total', 'selected']
python_df
```

```python
python_df['average'] = python_df['selected']/python_df['total']
python_df
```

```python
python_df.sort_values(by = ['total', 'selected'], ascending=False)
```

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/fasthill/My-gist/main/data/titanic/train.csv')
test_df = pd.read_csv('https://raw.githubusercontent.com/fasthill/My-gist/main/data/titanic/test.csv')
```

```python
train_df[['Pclass', 'Survived']].groupby(['Pclass']).mean().sort_values(by='Survived', ascending=False)
```

as_index: groupby를 사용하면 기본으로 그룹 라벨이 index로 설정됨. (index : Pclass)
Pclass를 index를 사용하고 싶은 않은 경우에는 as_index=False 를 설정하여 일련번호로 대체 가능

```python
train_df[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_df[['Sex', 'Survived']].groupby(['Sex'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_df[['SibSp', 'Survived']].groupby(['SibSp'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
train_df[['Parch', 'Survived']].groupby(['Parch'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```
