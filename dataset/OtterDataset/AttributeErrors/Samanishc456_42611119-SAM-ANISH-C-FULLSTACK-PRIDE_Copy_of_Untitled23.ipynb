```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with UN member data
data = {
    'Country': ['UN Members'],
    'Count': [193]
}

df = pd.DataFrame(data)

# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(df['Country'], df['Count'], color='blue')

# Add title and labels
plt.title('Number of United Nations Member States', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Members', fontsize=14)

# Show the count on top of the bar
for index, value in enumerate(df['Count']):
    plt.text(index, value, str(value), ha='center', va='bottom')

# Show the plot
plt.ylim(0, 200)  # Setting y-axis limit for better visibility
plt.show()
```
