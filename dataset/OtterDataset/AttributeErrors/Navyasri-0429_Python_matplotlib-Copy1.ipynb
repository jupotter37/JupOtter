```python
# area plot 
x = [1,2,3,5,6,7]
y = [10,14,8,16,12,13]
plt.fill_between(x,y, color = 'skyblue', alpha =0.6)
plt.plt(x,y, color = 'slateblue', alpha = 0.6, inewidth =2)
plt.title("Area Plot example")
plt.xlabel("X - aix")
plt.ylabel("Y-axis")
plt.grid(axis ='y')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data2.csv')
df
```

```python
plt.figure(figsize = (8,4))
plt.plot(df['Name'], df['Age'], marker = 'o', linestyle = '-', color = 'b')
plt.title("Line Plot - Name vs Age")
plt.xlabel("Name")
plt.ylabel("Age")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
# sactter plot
plt.figure(figsize = (8,4))
plt.scatter(df['Salary'], df['Rating'],marker = 'o', color = 'g', label = 'Data Points')
plt.title("Scatter Plot- Salary vs Rating ")
plt.xlabel("Salary"),
plt.ylabel("Rating")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# bar plot
plt.figure(figsize = (8,4))
plt.bar(df['Name'],df['Salary'], color='skyblue' , alpha = 0.6)
plt.title("Bar plot - Salary by Name ")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.grid(axis ='y')
plt.show()
```

```python
# histograph plot
plt.figure(figsize = (8,4))
plt.hist(df['Age'], bins = 3,color = 'purple', alpha = 0.7)
plt.title("Histogram - Age distribution") 
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis ='y')
plt.show()
```

```python
# pie chart
plt.figure(figsize = (6,6))
plt.pie( df['Salary'], labels = df['Name'], autopct = '%1.1f%%', startangle = 90)

#autopct = '%1.1f%%' ----> displays the percentage of each category on the pie chart with one decimal

plt.title('Pie chart - Salary Distribution')
plt.axis('equal') # equal ---> Equal aspect ratio that pie is drawn a circle
plt.show()
```

```python
plt.figure(figsize = (8,4))
plt.plot(df['Name'], df['Age'], marker = 'o', linestyle = '-', color = 'b')
plt.title("Line Plot - Name vs Age")
plt.xlabel("Name")
plt.ylabel("Age")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```
