```python
import numpy as np
import pandas as pd
```

```python
recipe_df = pd.read_csv('/Users/lisaheilscher/code/mariamorskovatykh/recipify/raw_data/recipes_and_interactions/RAW_recipes.csv')
recipe_df.head()
```

```python
#removing white spaces
recipe_df.columns = recipe_df.columns.str.replace(' ', '')
```

```python
#drop useless columns
recipe_df.drop(columns=['contributor_id', 'submitted', 'description'], inplace=True)
```

```python
#cleaning data
recipe_df['ingredients'] = recipe_df['ingredients'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
recipe_df['tags'] = recipe_df['tags'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
recipe_df['nutrition'] = recipe_df['nutrition'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
recipe_df['steps'] = recipe_df['steps'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
```

```python
#getting nutrition data in several columns(calories, sugar...)
recipe_df['calories'] = recipe_df['nutrition'].str.split(',').str[0]
recipe_df['total fat'] = recipe_df['nutrition'].str.split(',').str[1]
recipe_df['sugar'] = recipe_df['nutrition'].str.split(',').str[2]
recipe_df['sodium'] = recipe_df['nutrition'].str.split(',').str[3]
recipe_df['protein'] = recipe_df['nutrition'].str.split(',').str[4]
recipe_df['saturated fat'] = recipe_df['nutrition'].str.split(',').str[5]
recipe_df['carbohydrates'] = recipe_df['nutrition'].str.split(',').str[6]
```

```python
recipe_df.drop(columns=['nutrition'], inplace=True)
```

```python
recipe_df = recipe_df.rename(columns={'id': 'recipe_id'})
```
