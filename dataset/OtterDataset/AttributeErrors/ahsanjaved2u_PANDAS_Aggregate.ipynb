```python
#Practice of ratio of peapole having pytohn in each country
country_respondatn = df['Country'].value_counts()
country_respondatn
```

```python
country_group = df.groupby(['Country'])
country_respondant_with_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_respondant_with_python
```

```python
python_df = pd.concat([country_respondatn , country_respondant_with_python], axis='columns', sort =False)
python_df
```

```python
python_df.rename(columns={'count': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['NumRespondents']
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])* 100
python_df
```

```python

python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df
```
