```python
import pandas as pd
import geopandas as gpd
import os
```

```python
for e in os.listdir("."):
    if ".shp" in e:
        cur_file = gpd.read_file(e)
    else:
        continue
    cur_file.crs = "EPSG:4326"
    cur_file = cur_file.to_crs("EPSG:32614")
    cur_file.to_file("ProjectedRoutes/" + e)
```
