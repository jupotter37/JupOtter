### Problem 21
Mark each statement True or False. Justify the answer.

#### a.  A single vector by itself is linearly dependent.

__False__: Taking a set containing a single vector, that vector can never be a linear combination of any other vector in the set, since there are none.

#### b.  If $H = Span\{b_1, \dots, b_p\}$, then $\{b_1, \dots, b_p\}$ is a basis for H.

__False__: Only if $\{b_1, \dots, b_p\}$ is a linearly independent set of vectors.

#### c.  The columns of an invertible $n \times n$ matrix form a basis for $\Bbb R^n$.

__True__: By theorem 2.8, the invertible matrix theorem, the columns of an invertible $n \times n$ matrix are always linearly independant, and span $\Bbb R^n$, making them a basis for $\Bbb R^n$.

#### d.  A basis is a spanning set that is as large as possible.

__False__: A basis is a linearly independent set that is as large as possible, but a spanning set that is as small as possible.

#### e. In some cases, the linear dependence relations among the columns of a matrix can be affected by certain elementary row operations on the matrix.

__False__: Elementary row operations do not affect the  linear dependence relations of the columns.

### Problem 23
Suppose $\Bbb R^4 = Span \left\{v_1, \dots, v_4 \right\}$. Explain why $\left\{v_1, \dots, v_4 \right\}$ is a basis for $\Bbb R^4$.

By the definition of a subspace basis, the set must span the subspace, and it must be linearly independent.  
Since it is given that the set spans $\Bbb R^4$, it only remains to be shown that $\left\{v_1, \dots, v_4 \right\}$ must be linearly independent.  
Constructing a matrix from the vector set, will yield a $4 \times 4$ matrix. Now since the columns of the matrix span $\Bbb R^4$, the matrix is invertible, meaning that the columns of the matrix are linearly independent (The invertible matrix theorem 2.8 e and h)

### Problem 33
Consider the polynomials $\boldsymbol p_1 = 1 + t^2$ and $\boldsymbol p_2 = 1 - t^2$.  
Is $\{\boldsymbol p_1, \boldsymbol p_2\}$ linearly independent set in $\Bbb P^3$? Why or why not?

Firstly, this is a set in $\Bbb P^3$, since $\Bbb P^2$ is a subspace of $\Bbb P^3$.  
Secondly the set has to be linearly independent. To test this, one has to test if either $\boldsymbol p_1$ can be expressed as a linear combination of $\boldsymbol p_2$ or vice versa. Choosing to test the first one, the set is linearly independent iff no constant c exists such that:  
$\boldsymbol p_1 = c \boldsymbol p_2 \iff 1 + t^2 = c(1 - t^2) \iff c = \frac{1 + t^2}{1 - t^2}$  
which is clearly not a constant, and so __the set__ $\{\boldsymbol p_1, \boldsymbol p_2\}$ __is linearly independant.__
