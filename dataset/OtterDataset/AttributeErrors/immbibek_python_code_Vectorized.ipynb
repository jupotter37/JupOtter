```python
# problem in vectorized operations in vanilla python
s=['cat','mat',None,'rat'] # this list is caleed vector

[i.startswith('c') for i in s]
```

```python
# How pandas solves this issue?

s=pd.Series(['cat','mat',None,'rat'])
# string accesor
s.str.startswith('c')

# fast and optimized
```

```python
df=pd.read_csv('/content/titanic.csv')
df['Name']
```

```python
# Common Functions
# lower/upper/capitalized/title
df['Name'].str.upper()
# len
df['Name'][df['Name'].str.len()==82].values[0]

# strip
'                 nitish             '.strip()

```

```python
df['Name'].str.strip()
```

```python
# split --> get
df['Name'].str.split(',').str.get(0)     # if want sir name only i uesd get
```

```python
df[['title','firstname']]=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)

 # n=1 split only one
 # expand=True convert series inot Dataframe
```

```python
df.head(3)
```

```python
df['title'].value_counts()
```

```python
# replace
df['title'].str.replace('Ms','Miss')
df['title']=df['title'].str.replace('Mlle','Miss')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.startswith('A')]

# isdigit/isalpha
df[df['firstname'].str.isdigit()]
```

**Applying regular expression**

 contains

 search john--> both cases

 find lastnames with start and end char vowel

```python
df[df['firstname'].str.contains('john',case=False)]
```

```python
#find lastnames with start and end char vowel
df[df['firstname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
# here ^ this sighn represents starts and $ --> end
```
