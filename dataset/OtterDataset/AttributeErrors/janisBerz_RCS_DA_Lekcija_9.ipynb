```python
def preprocess_df(csv_file):
    df = pd.read_csv(csv_file,sep=';')
    # apstrādāsim datu tabulu
    # df = df[['Iepirkuma_ID', ' Iepirkuma_nosaukums',
    #    ' Iepirkuma_identifikacijas_numurs', ' Pasutitaja_nosaukums',
    #    ' Pasutitaja_registracijas_numurs',
    #    ' Pasutitaja_registracijas_numura_veids', ' Pasutitaja_PVS_ID',
    #    ' Augstak_stavosa_organizacija', ' Iepirkuma_statuss',
    #    ' Regulejasais_tiesibu_akts', ' Proceduras_veids',
    #    ' Hipersaite_EIS_kura_pieejams_zinojums',
    #    ' Hipersaite_uz_IUB_publikaciju', ' Ir_dalijums_dalas',
    #    ' Iepirkuma_dalas_nr', ' Iepirkuma_dalas_nosaukums',
    #    ' Iepirkuma_dalas_statuss', ' Uzvaretaja_nosaukums',
    #    ' Uzvaretaja_registracijas_numurs',
    #    ' Uzvaretaja_registracijas_numura_veids', ' Uzvaretaja_valsts',
    #    ' Liguma_dok_veids', ' Noslegta_liguma_dok_ID', ' Saistita_liguma_ID',
    #    ' Aktuala_liguma_summa', ' Aktuala_liguma_summas_valuta',
    #    ' Sakotneja_liguma_summa', ' Sakotneja_liguma_summas_valuta',
    #    ' Liguma_izpildes_termins', ' Liguma_izpilde_no',
    #    ' Liguma_izpilde_līdz', ' Liguma_dok_publicesanas_datums', ' Izbeigsanas_iemesls']]
    # izņemt liekas atstarpes
    df.columns = df.columns.str.strip()
    df = df.map(lambda x: x.replace('="','').replace('"','').strip())
    df = df.replace("",np.nan)
    df = df.astype({'Iepirkuma_ID':'int','Aktuala_liguma_summa':'float'})
    col_datumi = ['Liguma_izpilde_no','Liguma_izpilde_līdz','Liguma_dok_publicesanas_datums']
    for col in col_datumi:
        df[col] = pd.to_datetime(df[col],dayfirst=True,errors='coerce')
    df = df.drop_duplicates().reset_index(drop=True)
    return df
```
