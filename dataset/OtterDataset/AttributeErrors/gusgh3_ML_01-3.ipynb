```python
import pandas as pd
```

```python
df = pd.read_csv('data/Fish.csv')
df.head()
```

```python
df.shape
```

```python
df['Species'].unique()
```

```python
bream_df = df[df['Species'] == 'Bream']
# bream_df

bream_length = bream_df['Length2'].tolist()
bream_weight = bream_df['Weight'].tolist()

bream_length
# bream_weight
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight) #plt.scatter 함수: 산점도를 시각화(표현)해줌
plt.xlabel('length') # x축이 무엇을 나타내는 지 라벨로 보여줌
plt.ylabel('weight') # y축이 무엇을 나타내는 지 라벨로 보여줌
plt.show()
```

```python
smelt_df = df[df['Species'] == 'Smelt']
smelt_df.head()
```

```python
smelt_length = smelt_df['Length2'].tolist()
smelt_weight = smelt_df['Weight'].tolist()
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
#fish_data
```

```python
target_data = [1] * len(bream_length) + [0] * len(smelt_length)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, target_data)
```

```python
kn.score(fish_data, target_data)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(25, 400, c='black', s=100, marker='*') #하나의 점(특정 좌표에 맞는)을 찍고 싶을 때, 이처럼 사용. c=색깔, s=사이즈
plt.scatter(25, 150)
plt.scatter(smelt_length, smelt_weight)
plt.show
```

```python
kn.predict([[25, 400]])
```

```python
kn.predict([[16, 30]])
```

```python
kn.predict([[25, 150]])
```

```python
kn._fit_X #사이키런(머신러닝)이 학습한 내용을 알려줌
```

```python
kn._y #정답지: 뭐가 도미이고 smelt인지 알려줌
```

```python
kn49 = KNeighborsClassifier(49) #49개 데이터를 바라봐줘
```

```python
kn49.fit(fish_data, target_data) #fit:알고리즘에 학습시킬때 사용하는 함수
```

```python
kn49.score(fish_data, target_data)
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, target_data)

for n in range(5, 50):
    kn.n_neighbors = n

    score = kn.score(fish_data, target_data)
    
    print(n, score)
    if score < 1:
        break
```
