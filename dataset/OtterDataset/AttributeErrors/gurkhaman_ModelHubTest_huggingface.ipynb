```python
import yaml

# Function to convert YAML to a string
def yaml_to_string(yaml_content):
    try:
        # Load YAML content
        parsed_yaml = yaml.safe_load(yaml_content)
        # Convert back to string with your preferred format
        yaml_str = yaml.dump(parsed_yaml, default_flow_style=False)
        return yaml_str
    except Exception as e:
        return str(e)  # Returns the error message if conversion fails
    
```

```python
df['card_data'] = df['card_data'].apply(yaml_to_string)
```
