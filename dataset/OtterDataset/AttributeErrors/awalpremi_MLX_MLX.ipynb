### Fourier Transforms in MLX vs Numpy

##### Nice and Simple Numpy

```python
import numpy as np
import matplotlib.pyplot as plt

sampling_rate = 1000
duration = 1
t = np.linspace(0, duration, int(sampling_rate * duration),endpoint=False)

freq1 = 5
freq2 = 50
pi = np.pi
y1 = np.sin(2 * pi * freq1 * t)
y2 = np.sin(2 * pi * freq2 * t)
signal = y1 + y2

#Magic
fft = np.fft.fft(signal)
freq = np.fft.fftfreq(t.shape[-1],1/sampling_rate)


plt.subplot(2,1,1)
plt.plot(t,signal)
plt.subplot(2,1,2)
plt.stem(freq, np.abs(fft), 'b', markerfmt=" ", basefmt="-b")
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```
