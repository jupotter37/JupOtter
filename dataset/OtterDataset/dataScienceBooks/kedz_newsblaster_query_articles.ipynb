# Getting The Scope Out Of NewsBlaster

Researchers are often wasting precious time constantly obtaining the required corpus for their research. The same process is often repeated by  researcher after researcher (Graduate Students). 



###Find all articles that contain the word "language" in the *title* and return  only the *title* for the corresponding results###


```python
%rehash 
```

```python
curl -XGET 'island2.cs.columbia.edu:9200/news/_search?pretty&q=title:language*&fields=title'
```


###Find all articles that contain the word "innovative" in the "html content" and return both the "title" and "source link" for the corresponding results###


```python
curl -XGET 'island2.cs.columbia.edu:9200/news/_search?pretty&q=html_content:innovative*&fields=title,source_link'
```

###Same query as above but with additional meta information associated with that article###

```python
curl -XGET 'island2.cs.columbia.edu:9200/news/_search?pretty&q=html_content:innovative*&fields=title,meta_information.author,meta_information.topics'
```
