## Dia da semana com placares mais altos

Crie um gráfico que responda em qual dia da semana os placares foram mais altos. O conceito de "placar mais alto" é a soma dos gols feitos no jogo.

Como a quantidade de jogos não é a mesma em todos os dias da semana (finais de semana tendem a ter mais jogos que em dias da semana - o que resultaria em placares mais altos nestes dias), você deverá encontrar uma forma de equalizar esta diferença na quantidade de jogos.

A resposta deverá ser um gráfico que demonstre claramente o comparativo dos dias.

```python
# Responda aqui a etapa

df = matches_df

df['golsTotais'] = df['gf'] + df['ga']
mediaGolsPorDia = df.groupby('day')['golsTotais'].mean().reset_index()
print(mediaGolsPorDia)
diasDaSemana = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']

mediaGolsPorDia['day'] = pd.Categorical(mediaGolsPorDia['day'], categories=diasDaSemana, ordered=True)
mediaGolsPorDia = mediaGolsPorDia.sort_values('day')

fig = px.bar(mediaGolsPorDia, x='day', y='golsTotais', title='Média de gols por dia da semana')
fig.show()
```

## Formação Tática

## Classificação

Qual a classificação final do campeonato?

Responda em forma de tabela, ordenada do primeiro colocado ao último.
Você deverá apresentar o brasão do time (pode ser obtido online [aqui](https://logodetimes.com/premier-league/), mas automatize a busca, não salve o logo localmente), o nome, a quantidade de pontos e a posição final no campeonato.

**OBS** Vitória vale 3 pontos, empates 1 ponto e derrota 0 pontos.

```python
# Responda a etapa aqui

classificacao_df = matches_df.copy()
classificacao_df['pontos'] = classificacao_df['result'].apply(lambda x: 3 if x == 'W' else (1 if x == 'D' else 0))
# usa o pos para critério de desempate se equipe tiver jogo mais recente vai para frente
classificacao_df['pos'] = classificacao_df.groupby('date').cumcount() + 1
classificacao_df = classificacao_df.groupby('team').agg({'pontos': 'sum', 'pos': 'max'}).reset_index()

# ordena pelo total de pontos e pelo critério de desempate
classificacao_df = classificacao_df.sort_values(by=['pontos', 'pos'], ascending=[False, True])

classificacao_df['posicao'] = range(1, len(classificacao_df) + 1)

linkParaLogo = 'https://logodetimes.com/wp-content/uploads/'
for index, row in classificacao_df.iterrows():
    nomeTime = row['team']
    if nomeTime == 'Brighton and Hove Albion':
        nomeTime = 'Brighton hove albion'
    url = linkParaLogo + nomeTime.lower().replace(' ', '-') + '-football-club.png'
    response = requests.get(url)
    if response.status_code == 200:
        with open(f'{nomeTime}.png', 'wb') as f:
            f.write(response.content)
    else:
        print(f'Não foi possível baixar a imagem para {nomeTime}')
        # 3 times nao estão disponíveis no site fornecido pelo professor

classificacao_df['logo'] = classificacao_df['team'].apply(lambda x: f'<img src="https://logodetimes.com/wp-content/uploads/{x.lower().replace(" ", "-")}-football-club.png" alt="{x}" height="50">')

classificacao_df = classificacao_df.rename(columns={'team': 'Time', 'pontos': 'Pontos', 'posicao': 'Posição', 'logo' : 'Logo'})

display(HTML(classificacao_df[['Posição', 'Logo', 'Time', 'Pontos']].to_html(index=False, escape=False)))

```

## Público

Crie uma visualização que apresente o público total de cada equipe, e também um detalhamento por jogos.

Apresentar os dados usando a técnica "DrillDown" será considerado um "plus". A visualização DrillDown permite que gráficos sejam detalhados com cliques em itens do gráfico.

Um exemplo com tabelas pode ser encontrado [aqui](https://dash.plotly.com/dash-ag-grid/enterprise-master-detail).

Um exemplo usando gráficos pode ser encontrado [aqui](https://community.plotly.com/t/show-and-tell-drill-down-functionality-in-dash-using-callback-context/54403?u=atharvakatre)

```python
publicoTotal = matches_df.groupby('team')['attendance'].sum().reset_index()

publicoTotalOrdenado = publicoTotal.sort_values(by='attendance', ascending=False)
print(publicoTotalOrdenado)

fig_total = px.bar(publicoTotalOrdenado, x='team', y='attendance',
                   title='Público Total por Equipe')
fig_total.show()
```

```python
publicoPorJogo = matches_df.groupby(['team', 'date'])['attendance'].sum().reset_index()

def update_graph(timeSelecionado):
    import plotly.graph_objects as go
    filtered_df = publicoPorJogo[publicoPorJogo['team'] == timeSelecionado]

    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=filtered_df['date'],
        y=filtered_df['attendance'],
        name=timeSelecionado
    ))
    fig.update_layout(
        title=f'Público por Jogo - {timeSelecionado}',
        xaxis_title='Date',
        yaxis_title='Attendance',
        xaxis_tickangle=-45
    )
    fig.show()

team_dropdown = widgets.Dropdown(
    options=publicoPorJogo['team'].unique(),
    value=publicoPorJogo['team'].unique()[0],
    description='Time:',
)
# INFELIZMENTE para renderizar o grafico com dropdown é necessário rodar uma primeira vez antes do widget a função update graph
# Parece q o pyplot nao eh instanciado caso nao for feito isso
update_graph('')

widgets.interactive(update_graph, timeSelecionado=team_dropdown)
```
