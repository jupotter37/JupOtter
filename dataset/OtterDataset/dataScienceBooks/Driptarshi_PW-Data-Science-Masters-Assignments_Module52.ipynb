## Q1: Explain the following with an example:

*A) Artificial Intelligence (AI):*
Artificial Intelligence is the simulation of human intelligence processes by machines, especially computer systems. AI involves learning, reasoning, problem-solving, and perception.

Example: Virtual assistants like Siri or Alexa use AI to understand and respond to user queries.

*B) Machine Learning (ML):*
Machine Learning is a subset of AI that involves the use of algorithms and statistical models to enable a system to improve its performance on a task through experience.

Example: Email spam filtering, where the system learns to identify spam emails based on features from past emails.

*C) Deep Learning (DL):*
Deep Learning is a subset of ML that uses neural networks with many layers (deep networks) to model complex patterns in data.

Example: Image recognition systems that can identify objects in pictures with high accuracy, like Google's DeepDream.

## Q2: What is Supervised Learning? List some examples of supervised learning.

Supervised learning is a type of machine learning where the model is trained on labeled data, meaning the input data is paired with the correct output.

Examples:
- Linear regression
- Logistic regression
- Support Vector Machines (SVM)
- Decision trees
- Random forests
- Neural networks

## Q3: What is Unsupervised Learning? List some examples of unsupervised learning.

Unsupervised learning is a type of machine learning where the model is trained on data that has no labels and the system tries to learn the patterns and the structure from the data itself.

Examples:
- K-means clustering
- Hierarchical clustering
- Principal Component Analysis (PCA)
- Autoencoders

## Q4: What is the difference between AI, ML, DL, and DS?

- *AI (Artificial Intelligence):* The broadest concept that involves creating systems that simulate human intelligence.
- *ML (Machine Learning):* A subset of AI focused on systems that learn from data to improve their performance.
- *DL (Deep Learning):* A subset of ML that uses deep neural networks to model and understand complex patterns.
- *DS (Data Science):* An interdisciplinary field that uses scientific methods, processes, and algorithms to extract knowledge and insights from structured and unstructured data.

## Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

- *Supervised Learning:* Uses labeled data to train models, requiring a clear input-output pairing.
- *Unsupervised Learning:* Uses unlabeled data to find hidden patterns or intrinsic structures in the input data.
- *Semi-supervised Learning:* Uses a combination of a small amount of labeled data and a large amount of unlabeled data to improve learning accuracy.

## Q6: What is train, test, and validation split? Explain the importance of each term.

- *Train Set:* The portion of data used to train the model. It helps the model learn the patterns and relationships in the data.
- *Validation Set:* A separate portion of data used to tune the hyperparameters of the model and prevent overfitting. It helps in model selection.
- *Test Set:* The portion of data used to evaluate the final model's performance. It provides an unbiased evaluation of the model.

## Q7: How can unsupervised learning be used in anomaly detection?

Unsupervised learning can be used in anomaly detection by identifying patterns that do not conform to the expected behavior in the data. For example, clustering algorithms can group normal data points together, and any data points that do not fit well into these clusters can be considered anomalies.
