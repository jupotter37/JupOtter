## Neutral vs Polarised and Positive vs Negative Errors

Decomposition of errors into neutral vs polarised errors and positive vs negative errors.

```python
column_names = [column_name for column_name in list(df_phrase.columns) if column_name[0:6]=='count_']
column_names.append('tags')
column_names.append('text')

# test train split
X_train, X_test, y_train, y_test = train_test_split(df_phrase[column_names], 
                                                    df_phrase['sentiment'], 
                                                    test_size=config.TEST_SIZE, 
                                                    random_state=config.RANDOM_STATE)

# reset indicies -- required for heirachical classifier
# comparisson of models misclassified indicies in seperate notebook
# indicies required to be in line accross notebooks
X_train = X_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
def get_errors(y_pred, y_test):
    
    #
    n_np = np.logical_and(y_pred=='neutral', np.logical_or(y_test=='positive', y_test=='negative'))
    np_n = np.logical_and(np.logical_or(y_pred=='positive', y_pred=='negative'),y_test=='neutral')
    neutral_errors = np.logical_or(n_np, np_n)
    
    #
    n_p = np.logical_and(y_pred=='negative',y_test=='positive')
    p_n = np.logical_and(y_pred=='positive',y_test=='negative')
    poliarised_errors = np.logical_or(n_p, p_n)
    
    #
    assert((np.logical_or(neutral_errors, poliarised_errors) == (y_test != y_pred)).sum() == y_test.shape[0])
    
    #
    return neutral_errors, poliarised_errors
```

```python
df_expr = {'experiment':[], 'total error':[],  
           'error: neutral vs polarised':[], 'error: positive vs negative':[]}
for experiment in df_test.experiment:
    y_pred = np.array(ast.literal_eval(list(df_test[df_test.experiment == experiment].predictions)[0]))
    neutral_errors, polarised_errors = get_errors(y_pred, y_test)
    neutral_ratio, polarised_ratio= neutral_errors.sum()/y_test.shape[0], polarised_errors.sum()/y_test.shape[0]
    accuracy = list(df_test[df_test.experiment == experiment].accuarcy)[0]
    error = 1 - accuracy
    
    
    df_expr['experiment'].append(experiment)
    df_expr['total error'].append(error)
    df_expr['error: neutral vs polarised'].append(round(neutral_ratio,3))
    df_expr['error: positive vs negative'].append(round(polarised_ratio,3))
    
df_expr = pd.DataFrame(df_expr)
```

```python
cm = sns.light_palette("seagreen", as_cmap=True)
cm = cm.reversed()
styled_table = df_expr.iloc[[1,3,5]].style
styled_table.set_caption('Table 4. Total error decomposed into neutral vs polarised error and negative vs positive error. Total Error = 1 - Accuracy.')
styled_table
```

## Folder Structure for README.md

```python
import os

def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print('{}{}/'.format(indent, os.path.basename(root)))
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print('{}{}'.format(subindent, f))
```
