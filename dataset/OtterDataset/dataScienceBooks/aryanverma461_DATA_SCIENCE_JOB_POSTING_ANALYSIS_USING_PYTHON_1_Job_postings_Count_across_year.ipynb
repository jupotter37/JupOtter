```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
parquet_file= r'C:\Users\verma\OneDrive\Desktop\PYTHON_DATA_PROJECT\1_Data_science_job_postings_dataset\0000.parquet'
df = pd.read_parquet(parquet_file, engine='auto')

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
job_counts = df['job_posted_month'].value_counts()
job_counts.sort_index(inplace=True)
```

```python
job_counts.plot(kind='line', marker='o')
ax =plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _:f'${int(x/1000)}K'))
plt.title('Job Postings Across Year')
plt.ylabel('Count of job')
plt.xlabel('Month')
plt.xticks(ticks=range(0, 13, 2))
plt.show()
```
