```python
# Warmup 0: Recall how to read in the csv data

import csv

# source:  Automate the Boring Stuff with Python
def process_csv(filename):
    exampleFile = open(filename, encoding="utf-8")  
    exampleReader = csv.reader(exampleFile) 
    exampleData = list(exampleReader)        
    exampleFile.close()  
    return exampleData

survey_data = process_csv('cs220_survey_data.csv')
header = survey_data[0]
header
```

```python
# Warmup 1: put survey_data into buckets by lecture
# make a dictionary of lists
# key is the lecture

rows = survey_data[1:]

lecture_dict = {}
for row in rows:
    lecture = row[header.index('lecture')]
    age = row[header.index('age')]
    #print (lecture, age)
    if lecture not in lecture_dict: # first time
        lecture_dict[lecture] = [] # make an empty list
    lecture_dict[lecture].append(row)
# lecture_dict['LEC003']
```

```python
# Warmup 2: Find the average age for our lecture

sum_of_ages = 0
lecture_name = "LEC003"
num_people = 0
#print(type(lecture_dict[lecture_name]))
for row in lecture_dict[lecture_name]:
    age = row[header.index('age')]
    if age == '':
        continue
    sum_of_ages += int(age)
    num_people += 1
average = round(sum_of_ages/num_people, 3)
print(lecture_name, average)
```

```python
# Warmup 3: Make a dictionary of all average ages
# the key is the lecture name
# the value is the average age

# tabbing the copied code
# first just print out all the results,
# then add to a dictionary
lec_age_dict = {}
for key in lecture_dict:
    #print(key)
    sum_of_ages = 0
    lecture_name = key
    num_people = 0
    for row in lecture_dict[lecture_name]:
        age = row[header.index('age')]
        if age == '':
            continue
        sum_of_ages += int(age)
        num_people += 1

    average = round(sum_of_ages/num_people, 3)
    print(lecture_name, average)
    # now put into a dictionary
    lec_age_dict[lecture_name] = average
lec_age_dict
```

```python
# Let's practice reading in a sample JSON file
import json

def read_json(path):
    with open(path, encoding="utf-8") as f: # f is a varaible 
        return json.load(f)                 # f represents a reference the JSON file
    
def write_json(path, data):
    with open(path, 'w', encoding="utf-8") as f:
        json.dump(data, f, indent=2)
```

```python
# first, let's take a look at the file score_history.json
```
