```python
from boxoffice.db.frames import get_movie_frame, get_box_office_day_frame

movie_frame = get_movie_frame()

box_office_day_frame = get_box_office_day_frame()
```

```python
# want to add the total box office for each movie to the movie frame
sums = box_office_day_frame.groupby('movie')['revenue'].sum()

movie_frame['total_box_office'] = movie_frame['id'].map(sums)

movie_frame.head()
```

```python
# average or median box office per value
median_by_production_method = movie_frame.groupby('production_method')['total_box_office'].median().sort_values(ascending=False)

print(median_by_production_method)

# now do it for creative type
median_by_creative_type = movie_frame.groupby('creative_type')['total_box_office'].median().sort_values(ascending=False)

print(median_by_creative_type)
```
