```python
# Carga el dataset de trabajo
import pandas as pd
pd.set_option("display.max_columns",None)
df = pd.read_csv("datos/netflix_mergeado.csv", index_col=0)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
movie = df["type"] == "Movie"
lan2021 = df["release_year"] == 2021
pelis_2021 = df[movie & lan2021]
pelis_2021["title"]
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
mins_90 = df.loc[df["duration"] == "90 min", ["title","duration"]]
mins_90
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
ano = df["release_year"] >= 2018
pais =  df["country"] == "India"
india_2018 = df.loc[pais & ano, ["title"]]
india_2018
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
tv_14 = df["rating"] == "TV-14"
r =  df["rating"] == "R"
calificacion = df.loc[tv_14 | r, ["title"]]
calificacion
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
nolan = df["director"] == "Christopher Nolan"
scorsese = df["director"] == "Martin Scorsese"
dirigidos_ambos = df.loc[nolan | scorsese, ["title", "director"]]
print((f"Los títulos que han dirigido son {dirigidos_ambos}"))
print((f"Los títulos dirigidos por Nolan son {dirigidos_ambos[nolan].value_counts("title").sum()}"))
print((f"Los títulos dirigidos por Scorsese son {dirigidos_ambos[scorsese].value_counts("title").sum()}"))
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
usa =  df["country"] == "United States"
uk =  df["country"] == "United Kingdom"
canada =  df["country"] == "Canada"
lista = df.loc[usa | uk | canada, ["title"]]
patron_masidiomas = ".*/*"
masidiomas = df[df["Language"].str.contains(patron_masidiomas, regex = True, na = False)]
print ((f"Los títulos que pertenecen a USA, UK o Canada son {lista}"))
print((f"El numero de películas con mas de un lenguage es de: {masidiomas.loc[usa | uk | canada].value_counts("title").sum()}"))
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
df_2015a2020 = df[df["release_year"].between(2015, 2020, inclusive = "both" )]
df_2015a2020["title"]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
df_imdb= df[df["IMDB Score"].between(6.5, 8.0, inclusive = "both" )]
df_imdb = df.loc[df["duration"].isnull() == True, ["title"]]
df_imdb
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
pelis90 = df["duration"] == "90 min"
pelis150 =  df["rating"] == "150 min"
pelis90y150 = df.loc[pelis90 | pelis150, ["title"]]
pelis90y150
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
df_tvshow = df[df["type"] == "TV Show"]
patron = ".*love*"

df_love = df[df["title"].str.contains(patron, regex = True, na = False, case = False)]
df_love.title
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron = ".*Johnson*"

df_johnson = df[df["director"].str.contains(patron, regex = True, na = False, case = False)]
df_johnson.loc[:,["title","director"]]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python
patron = ".*Documentaries*"
df_docu = df[df["listed_in"].str.contains(patron, regex = True, na = True, case = False)]
patron1 = ".*Mike Tyson.*"
df_tyson = df_docu[df_docu["description"].str.contains(patron1, regex = True, na = True, case = True)]
df_tyson.loc[:,["title"]]
```
