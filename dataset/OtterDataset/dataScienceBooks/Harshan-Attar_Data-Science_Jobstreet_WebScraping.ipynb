### Problem Statement : Scrape The Data From JobStreet And Naukri.com.  Save the jobs info in csv files. You will create one csv file for JobStreet and one for Naukri.com.

```python
# Importing Necessary libraries
import pandas as pd
import numpy as np
import time
import requests
import selenium
from selenium import webdriver
import warnings
import os
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup


# Scraping The Data From JobStreet
def getUrl(key):
    key = key.replace(' ','-')
    temp = 'https://www.jobstreet.com.my/{}-jobs'
    url = temp.format(key)
    return url

def find_job(card):
    # Scraping job_name
    try:
        job_name = card.find('h3','y735df0 _1iz8dgs4y _94v4w0 _94v4wl _1wzghjf4 _94v4wv _94v4w21').text
    except:
        job_name = 'Not Mentioned'
        
    # Scraping company_name
    try:
        company_name = card.find('span','y735df0 _1iz8dgs4y _94v4w0 _94v4w1 _94v4w21 _1wzghjf4 _94v4wa').text
    except:
        company_name = 'Not Mentioned'

    # Scraping company_location
    try:
        company_location = card.find('span','y735df0 _1iz8dgs4y _94v4w0 _94v4w1 _94v4w21 _1wzghjf4 _94v4w7').text
    except:
        company_location = 'Not Mentioned'
        
    # Scraping salary
    try:
        salary = card.find('span','y735df0 _153p76c1 _1iz8dgs4y _1iz8dgs0 _1iz8dgsr _153p76c3').text
    except:
        salary = 'Not Mentioned'
    
    # Scraping job_description
    try:
        job_description = card.find('ul','y735df0 y735df3 _1akoxc50 _1akoxc54').text
    except:
        job_description = 'Not Mentioned'
        
    
    # Scraping job_apply_url
    try:
        job_apply_url = card.find('h3','y735df0 _1iz8dgs4y _94v4w0 _94v4wl _1wzghjf4 _94v4wv _94v4w21').find('a')
        job_apply_url = 'https://www.jobstreet.com.my' + job_apply_url['href']
    except:
        job_apply_url = 'Not Mentioned'

    # Scraping job posting date
    try:
        job_posted = card.find('span','y735df0 _1iz8dgs4y _94v4w0 _94v4w1 _94v4w22 _1wzghjf4 _94v4w7').text
    except:
        job_posted = 'Not Mentioned'

    all_job_detail = (job_name,company_name,company_location,salary,job_description,job_posted,job_apply_url)
    return all_job_detail

def main(key):
    all_jobs = []
    url = getUrl(key)
    res = requests.get(url)
    soup = BeautifulSoup(res.content,'html.parser')
    all_divs = soup.find_all('article','y735df0 y735df1 _1iz8dgs7i _1iz8dgs6e _1iz8dgs9q _1iz8dgs8m _1iz8dgs66 _1iz8dgsh _1iz8dgs5e uo6mkb _1iz8dgs4i _1iz8dgs4b _94v4w18 _94v4w1b _1iz8dgs32 _1iz8dgs35')
    for i in all_divs:
        temp = find_job(i)
        all_jobs.append(temp)
    col = ['Job Name','Company','Location','Salary','Job Description','Job Posted','Apply url']
    jobstreet_jobs_data = pd.DataFrame(all_jobs,columns = col)
    dir_path = r'C:\Users\Harshan Attar\Desktop\iGurus Practice\jobs'
    file_name = 'jobstreetjobs.csv'

    file_path = os.path.join(dir_path, file_name)
    jobstreet_jobs_data.to_csv(file_path,encoding='utf-8')
```

```python
main('Java Dev')
```
