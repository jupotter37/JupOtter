```python
def mmr_train():
    f = open('mmr3.txt','r')
    content = f.read().splitlines()
    f.close()
    fw = open('feature.txt', 'w')
    fr = open('feature.txt', 'r')
    fcontent = fr.read().splitlines()
    x_train, y_train = gen_vector(fcontent)
    norm_txt(fw,content)
    fw.close()
    return x_train, y_train
    
    
```

```python
def norm_txt(fw,content):
    import nltk
    from nltk.stem import WordNetLemmatizer
    for sen in content:
        sensplit = sen.split()
        label = sensplit[-1]  # last word
        sen = sensplit[0:len(sensplit)-1]
        sen = ' '.join(sen)   # joining the splitted words
        nouns = []
        nouns.append(label)
        words = nltk.word_tokenize(str(sen))
        for word, pos in nltk.pos_tag(words):      
            if(pos == 'NN' or pos =='NNP' or pos =='NNS' or pos =='NNPS'):
                word = WordNetLemmatizer().lemmatize(word)
                nouns.append(word)
        
        fea = ' '.join(nouns)
        fw.write(fea)
        fw.write('\n')
        
            
```

```python
def gen_bow():    # generates unique words
    import nltk
    fw = open('feature.txt','r')
    content = fw.read().splitlines()
    bow = []
    for sen in content:
        sent = sen.split()
        sen  = sent[1:len(sent)]
        sen = ' '.join(sen)
        words = nltk.word_tokenize(str(sen))
        for word in words:
            bow.append(word)
    bow = list(set(bow))
    print(' ')
    print(len(bow))
    return bow
    
```

<h1>Feature Extraction </h1>

```python
def gen_vector(fcontent):
    import nltk
    bow = gen_bow()
    x = []
    y = []
    for sentence in fcontent:
        fea_vector = []
        lab = []
        for i in bow:
            fea_vector.append(0)
        label = sentence.split(' ',1)[0]
        fea = sentence.split(' ',1)[1]
        if label == 'relevant':
            l = 1
        if label == 'irrelevant':    
            l = 0
        lab.append(l)
        token = nltk.word_tokenize(fea)    
        for word in token:
            if word not in bow:
                continue
            pos = bow.index(word)
            fea_vector[pos] = 1
        x.append(fea_vector)
        y.append(lab)
    return x,y        
        
```

```python
def simple_nn():
    import numpy as np
    import keras
    from keras.models import Sequential
    from keras.layers.core import Dense
    x_train, y_train = mmr_train()
    x_train = np.array(x_train)
    y_train = np.array(y_train)
    model = Sequential()
    model.add(Dense(64, input_dim = 567, activation = 'relu'))
    model.add(Dense(32, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
    model.fit(x_train, y_train, epochs = 10, verbose = 2)
    return model
```
