## List and describe the different types of databases

    1) Single-user
        *database that supports one user at a time
        
    2) Desktop
        * Single-user database that runs on a personal computer

    3) Multi-user
        * Database that supports multiple users at the same time
        
    4) Workgroup
        * Multi-user database that supports a single department or a small number of users
        
    5) Enterprise
        * Multi-user database that supports an organization or a large group of users
