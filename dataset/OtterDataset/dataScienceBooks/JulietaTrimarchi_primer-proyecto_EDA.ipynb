```python
# creo una función para detectar cadenas vacías
def is_empty_string(value:str)->bool:

    """Verifica si el valor proporcionado es una cadena vacía.


    Args:
        value (str): El valor que se va a verificar.

    Returns:
        bool: True si el valor es una cadena vacía (después de eliminar espacios en blanco alrededor),de lo contrario, False.
    """
    
    return isinstance(value, str) and value.strip() == ''

# Creo una máscara booleana para filas con cadenas vacías
empty_string_mask_genres = data_total['genres_names'].apply(is_empty_string)

# Filtro las filas con cadenas vacías
rows_with_empty_strings_genres = data_total[empty_string_mask_genres]

# Muestro las filas con cadenas vacías
print("Filas con cadenas vacías en 'genres_names':")
rows_with_empty_strings_genres
```

```python
# Voy a transformar los datos de la columna 'popularity' de tipo str a tipo float
data_total['popularity'] = pd.to_numeric(data_total['popularity'], errors='coerce')

# Verifico
# Tipo de datos
print(data_total['popularity'].apply(type).value_counts())
```

```python
# A partir del análisis realizado voy a realizar los recortes que considero necesarios para el posterior modelo de machine learning

# Elimino filas con nulos encontradas en columnas importantes como 'original_languages','overview','genres_names','actors_names','directors_names'

# Mascaras creadas anteriormente:
null_mask_languages = data_total['original_language'].isna()
null_mask_overview = data_total['overview'].isna()
empty_string_mask_genres = data_total['genres_names'].apply(is_empty_string)
empty_string_mask_directors = data_total['directors_names'].apply(is_empty_string)
empty_string_mask_actors = data_total['actors_names'].apply(is_empty_string)

combined_mask = (
                    null_mask_languages  |
                    null_mask_overview |
                    empty_string_mask_genres |
                    empty_string_mask_directors |
                    empty_string_mask_actors 
)

# Filtro el DataFrame para conservar solo las filas que no están marcadas para eliminar
data_filtered = data_total[~combined_mask]
data_filtered
```

```python
# Cuento las frecuencias de cada valor en la columna 'original_language'
value_counts = data_filtered['original_language'].value_counts()

# Filtro y cuento solo los valores 'es' y 'en'
count_es_en = value_counts.get('es', 0) + value_counts.get('en', 0)

print(f"Cantidad de filas donde 'original_language' es 'es' o 'en': {count_es_en}")
```

```python
# Realizaré un recorte y dejaré solo las peliculas en las que el original_language es español o ingles

languages_of_interest = ['es', 'en']
data_filtered = data_filtered[data_filtered['original_language'].isin(languages_of_interest)]
data_filtered
```

```python
data = data_filtered.copy()
```

```python
# Voy a filtrar el dataframe por un rango de popularidad y rango de años de estreno más recientes

# Defino los rangos basado en el anpalisis anterior
popularity_min = 0.55
popularity_max = 5.39
year_min = 2000
year_max = 2020

# Filtro por 'popularity'
data= data[(data['popularity'] >= popularity_min) & (data['popularity'] <= popularity_max)]

# Filtra por años
data = data[(data['release_year'] >= year_min) & (data['release_year'] <= year_max)]

# Opcional: reiniciar el índice después del filtrado
data.reset_index(drop=True, inplace=True)

data
```

```python
# Analizo peliculas duplicadas

mask_duplicados = data['title'].duplicated(keep=False)
num_duplicados = mask_duplicados.sum()
print(num_duplicados)

df_duplicados = data[mask_duplicados]
result = df_duplicados.groupby('title',group_keys=False).apply(lambda x: list(x.index)).reset_index(name='Indices')
print(result)
```
