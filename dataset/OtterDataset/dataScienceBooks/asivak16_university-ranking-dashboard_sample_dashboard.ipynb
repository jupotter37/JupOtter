```python
import pandas as pd
import plotly.express as px
from dash import dcc, html
from dash import Dash
from jupyter_dash import JupyterDash
```

```python
df = pd.read_csv("../data/2022_rankings_cleaned.csv")
df.head()
```

```python
app = Dash(__name__)

# Create scatter plot for Citations vs. World Rank
fig_scatter = px.scatter(df, x='rank', y='scores_citations', color='location',
                        title='Citations vs. World Rank by Location')

# Create histogram for Distribution of World Rankings
fig_histogram = px.histogram(df, x='rank', nbins=50, title='Distribution of World Rankings',
                             labels={'rank': 'World Rank'})

# Define layout of the app
app.layout = html.Div(children=[
    html.H1(children='University Rankings Dashboard'),

    html.Div(children='''
        A simple dashboard visualizing university rankings data.
    '''),

    dcc.Graph(
        id='citations-vs-rank',
        figure=fig_scatter
    ),

    dcc.Graph(
        id='distribution-of-world-rank',
        figure=fig_histogram
    )
])

# Run the app
if __name__ == '__main__':
    app.run_server(mode='inline', debug=True)
```
