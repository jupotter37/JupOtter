```python
#Discuss string slicing and provie examples.

#String slicing is a powerful feature in Python that allows you to extract parts of a string using a specific syntax. Let’s break it down and look at some examples.

#1 Examples- Extracting a Substring
text = "Hello, World!"
print(text[0:5])  # Output: Hello
print(text[7:12]) # Output: World

#2 Slicing with Steps
text = "Hello, World!"
print(text[::2])  # Output: Hlo ol!
print(text[1::2]) # Output: el,Wrd

#3Negative Indexing
text = "Hello, World!"
print(text[-6:-1]) # Output: World
print(text[::-1])  # Output: !dlroW ,olleH (reverses the string)

#4Omitting Start or Stop
text = "Hello, World!"
print(text[:5])  # Output: Hello (from start to index 4)
print(text[7:])  # Output: World! (from index 7 to end)






```

```python
#Explain key features of list in Python.

#Lists in Python are one of the most versatile and commonly used data structures. Here are some key features:

#1. Ordered- Lists maintain the order of elements. This means that the items have a defined sequence, and this order will not change unless explicitly modified.

#2 Mutable- Lists are mutable, meaning you can change, add, or remove items after the list has been created.
my_list = [1, 2, 3]
my_list[0] = 10  # Changes the first element to 10
print(my_list)  # Output: [10, 2, 3]

#3. Allow Duplicates- Lists can contain duplicate elements. This is useful when you need to store multiple instances of the same value.

#4 Dynamic Size- Lists can grow and shrink dynamically as you add or remove elements.
my_list = [1, 2, 3]
my_list.append(4)  # Adds 4 to the end of the list
print(my_list)  # Output: [1, 2, 3, 4]

#5. Heterogeneous Elements- Lists can store elements of different data types, including integers, strings, and even other lists.
my_list = [1, "Hello", 3.14, [1, 2, 3]]
print(my_list)  # Output: [1, 'Hello', 3.14, [1, 2, 3]]


```

```python
#Describe how to access, modify and delete elements in a list with examples.

#Accessing Elements
#You can access elements in a list using indexing. Python uses zero-based indexing, so the first element is at index 0.

#Examples- 
my_list = [10, 20, 30, 40, 50]
print(my_list[0])  # Output: 10
print(my_list[2])  # Output: 30
print(my_list[-1]) # Output: 50 (last element)

#Modifying Elements
#Lists are mutable, meaning you can change their elements.
my_list = [10, 20, 30, 40, 50]
my_list[1] = 25
print(my_list)  # Output: [10, 25, 30, 40, 50]

#Deleting Elements
#You can delete elements from a list using the del statement, remove() method, or pop() method.
my_list = [10, 20, 30, 40, 50]
del my_list[1]
print(my_list)  # Output: [10, 30, 40, 50]

#Using remove():
my_list = [10, 20, 30, 40, 50]
my_list.remove(30)
print(my_list)  # Output: [10, 20, 40, 50]

#Using pop():
my_list = [10, 20, 30, 40, 50]
popped_element = my_list.pop(2)
print(my_list)  # Output: [10, 20, 40, 50]
print(popped_element)  # Output: 30

#Summary
#Accessing: Use indexing (e.g., my_list[0]).
#Modifying: Assign a new value to an index (e.g., my_list[1] = 25).
#Deleting: Use del, remove(), or pop().


```

```python
#Compare and contrast tuples and list with examples.

#Lists
#Mutable: You can modify lists after their creation.
#Dynamic Size: Lists can grow and shrink as needed.
#Syntax: Defined using square brackets [].

my_list = [1, 2, 3, 4]
my_list[1] = 20  # Modifying an element
my_list.append(5)  # Adding an element
print(my_list)  # Output: [1, 20, 3, 4, 5]

#Tuples
#Immutable: Once created, you cannot modify tuples.
#Fixed Size: The size of a tuple is fixed after creation.
#Syntax: Defined using parentheses ().

my_tuple = (1, 2, 3, 4)
# my_tuple[1] = 20  # This will raise an error
print(my_tuple)  # Output: (1, 2, 3, 4)

#Key Differences

#Mutability:

#List: Mutable (can be changed).

#Tuple: Immutable (cannot be changed).

#Syntax:

#List: []
#Tuple: ()

#Performance:
#List: Slightly slower due to dynamic nature.
#Tuple: Faster due to immutability.

#Methods:

#List: Many built-in methods like append(), remove(), sort(), etc.
#Tuple: Fewer methods, mainly count() and index().

#Use Cases:
#List: Suitable for collections of items that may change.
#Tuple: Suitable for fixed collections of items.

#Examples of Tuple methods-

my_tuple = (1, 2, 3, 4, 2)
print(my_tuple.count(2))  # Output: 2
print(my_tuple.index(3))  # Output: 2
```

```python
#Define the key features of sets and provide examples of their use.

#Key Features of Sets

#1- Unordered Collection
#Sets do not maintain any particular order of elements.
my_set = {3, 1, 4, 2}
print(my_set)  # Output: {1, 2, 3, 4} (order may vary)

#2- Unique Elements
#Sets automatically remove duplicate elements.
my_set = {1, 2, 2, 3, 4, 4}
print(my_set)  # Output: {1, 2, 3, 4}

#3- Mutable
#You can add or remove elements from a set.
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}
my_set.remove(2)
print(my_set)  # Output: {1, 3, 4}

#4- No Indexing
#Sets do not support indexing or slicing.

my_set = {1, 2, 3}
# my_set[0]  # This will raise an error


#5- Efficient Membership Testing
#Checking if an element is in a set is very fast.
my_set = {1, 2, 3}
print(2 in my_set)  # Output: True
print(5 in my_set)  # Output: False

#Use Cases
#Removing Duplicates: Quickly remove duplicates from a list by converting it to a set.
my_list = [1, 2, 2, 3, 4, 4]
my_set = set(my_list)
print(my_set)  # Output: {1, 2, 3, 4}

#Membership Testing: Efficiently check if an item exists in a collection.
my_set = {1, 2, 3, 4}
print(3 in my_set)  # Output: True

#Set Operations: Perform mathematical set operations like union, intersection, etc.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1 | set2)  # Output: {1, 2, 3, 4, 5}









```

```python
#Discuss use cases of tuples and sets in python programming.

#Use Cases of Tuples

#1- Immutable Data Storage

#2- Configuration Settings: Tuples are often used to store configuration settings that should not change throughout the program.

config = ("localhost", 8080, "admin", "password")

#Returning Multiple Values from Functions
#Function Return Values: Functions can return multiple values as a tuple.
def get_user_info():
    return ("John", "Doe", 30)

first_name, last_name, age = get_user_info()

#3- Dictionary Keys
#Composite Keys: Tuples can be used as keys in dictionaries because they are immutable.
coordinates = {(0, 0): "Origin", (1, 2): "Point A"}

#4- Data Integrity
#Ensuring Data Integrity: Tuples ensure that the data remains unchanged, which is useful for fixed collections of items.
months = ("January", "February", "March", "April")

#5- Efficient Iteration
#Iteration: Tuples can be more memory-efficient and faster to iterate over compared to lists.
for month in ("January", "February", "March"):
    print(month)
    
#Use Cases of Sets

#1-Removing Duplicates

#Unique Elements: Sets automatically remove duplicate elements, making them ideal for filtering unique items from a list.

my_list = [1, 2, 2, 3, 4, 4]
unique_set = set(my_list)
print(unique_set)  # Output: {1, 2, 3, 4}

#2-Membership Testing

#Fast Membership Testing: Sets provide an efficient way to check for the presence of an element.

my_set = {1, 2, 3, 4}
print(3 in my_set)  # Output: True

#3-Set Operations

#Mathematical Operations: Sets support operations like union, intersection, difference, and symmetric difference.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1 | set2
print(union_set)  # Output: {1, 2, 3, 4, 5}


#4-Data Validation

#Validation: Sets can be used to validate data by ensuring that only unique items are present.

valid_ids = {101, 102, 103}
user_id = 104
if user_id in valid_ids:
    print("Valid ID")
else:
    print("Invalid ID")

#5-Efficient Data Storage
#Storing Unique Items: Sets are useful for storing unique items without any particular order.

unique_words = {"apple", "banana", "cherry"}

#Summary

#Tuples are ideal for immutable data storage, returning multiple values from functions, using as dictionary keys, ensuring data integrity, and efficient iteration.

#Sets are perfect for removing duplicates, fast membership testing, performing set operations, data validation, and storing unique items.
```

```python
#Describe how to add modify and delete items in a dictionary with examples

#Adding Items
#You can add items to a dictionary by assigning a value to a new key.

#Example:

my_dict = {"name": "Alice", "age": 25}
my_dict["city"] = "New York"
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

#Modifying Items

#You can modify the value of an existing key by reassigning it.

#Example:

my_dict = {"name": "Alice", "age": 25}
my_dict["age"] = 26
print(my_dict)  # Output: {'name': 'Alice', 'age': 26}

#Deleting Items
#You can delete items from a dictionary using the del statement, pop() method, or popitem() method.

my_dict = {"name": "Alice", "age": 25, "city": "New York"}
del my_dict["city"]
print(my_dict)  # Output: {'name': 'Alice', 'age': 25}

#Using pop():

my_dict = {"name": "Alice", "age": 25, "city": "New York"}
age = my_dict.pop("age")
print(my_dict)  # Output: {'name': 'Alice', 'city': 'New York'}
print(age)      # Output: 25

#Using popitem():
    
#This method removes and returns the last inserted key-value pair.

my_dict = {"name": "Alice", "age": 25, "city": "New York"}
item = my_dict.popitem()
print(my_dict)  # Output: {'name': 'Alice', 'age': 25}
print(item)     # Output: ('city', 'New York')

#Summary

#Adding: Use assignment (e.g., my_dict["new_key"] = value).

#Modifying: Reassign the value to an existing key (e.g., my_dict["key"] = new_value).

#Deleting: Use del, pop(), or popitem().
```

```python
#Discuss the importance of dictionary keys being immutable and provide examples.

#Importance of Immutable Dictionary Keys

#1- Hashability

#Hash Function: Dictionaries use a hash table to store key-value pairs. The hash function generates a unique hash value for each key, which is used to quickly locate the corresponding value.

#Consistency: If a key were mutable, its hash value could change, making it impossible to reliably retrieve the associated value12.

#2-Performance

#Efficiency: Immutable keys ensure that the hash value remains constant, allowing for efficient lookups, insertions, and deletions.

#Predictability: Mutable keys could lead to unpredictable behavior and degraded performance due to the need to rehash keys12.

#3-Data Integrity

#Consistency: Immutable keys prevent accidental changes that could corrupt the dictionary’s structure.

#Reliability: Ensures that the key-value mapping remains consistent throughout the program’s execution12.

#Examples of Immutable Keys

#Using Strings as Keys
my_dict = {"name": "Alice", "age": 25}
print(my_dict["name"])  

#Using Tuples as Keys
my_dict = {(1, 2): "point A", (3, 4): "point B"}
print(my_dict[(1, 2)])  

#Example of Mutable Keys (Incorrect Usage)

#Using a list as a key will raise an error because lists are mutable:
my_dict = {[1, 2]: "value"} 

#Converting Mutable to Immutable

#If you need to use a mutable type as a key, convert it to an immutable type:
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
my_dict = {my_tuple: "value"}
print(my_dict[(1, 2, 3)]) 

#Summary

#Immutable keys ensure the reliability and efficiency of dictionary operations.

#Common immutable types used as keys include strings, numbers, and tuples.
```
