**Problem2-Leetcode#1-Two Sum-Easy**

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

- Example 1:
    - Input: nums = [2,7,11,15], target = 9
    - Output: [0,1]
    - Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

- Example 2:
    - Input: nums = [3,2,4], target = 6
    - Output: [1,2]

- Example 3:
    - Input: nums = [3,3], target = 6
    - Output: [0,1]


Here’s how it works:

- So I will create an empty dictionary mapEle.
- It then loops over the nums list. For each number, it checks if target - nums[i] is in mapEle. If it is, that means it has found two numbers that add up to target, and it returns their indices.
- If target - nums[i] is not in mapEle, it adds the current number and its index to mapEle. This is done so that it can be checked against in future iterations.
- If no two numbers add up to target, it returns None.

As for the time and space complexity:

Time Complexity: The time complexity is O(n), where n is the length of the nums list. This is because it loops through the nums list once.

Space Complexity: The space complexity is also O(n), where n is the length of the nums list. This is because in the worst-case scenario, it might end up storing every number from nums in the mapEle dictionary.

**Problem3-Leetcode#146-LRU Cache-Medium**

Design a data structure that follows the constraints of a Least Recently Used (LRU) cache.

Implement the LRUCache class:

LRUCache(int capacity) Initialize the LRU cache with positive size capacity.\
int get(int key) Return the value of the key if the key exists, otherwise return -1.\
void put(int key, int value) Update the value of the key if the key exists. Otherwise,\
add the key-value pair to the cache. If the number of keys exceeds the capacity from \
this operation, evict the least recently used key.

The functions get and put must each run in O(1) average time complexity.

 

- Example 1:

    - Input ["LRUCache", "put", "put", "get", "put", "get", "put", "get", "get", "get"]\
    [[2], [1, 1], [2, 2], [1], [3, 3], [2], [4, 4], [1], [3], [4]]
    - Output
    [null, null, null, 1, null, -1, null, -1, 3, 4]

    - Explanation
        - LRUCache lRUCache = new LRUCache(2);
        - lRUCache.put(1, 1); // cache is {1=1}
        - lRUCache.put(2, 2); // cache is {1=1, 2=2}
        - lRUCache.get(1);    // return 1
        - lRUCache.put(3, 3); // LRU key was 2, evicts key 2, cache is {1=1, 3=3}
        - lRUCache.get(2);    // returns -1 (not found)
        - lRUCache.put(4, 4); // LRU key was 1, evicts key 1, cache is {4=4, 3=3}
        - lRUCache.get(1);    // return -1 (not found)
        - lRUCache.get(3);    // return 3
        - lRUCache.get(4);    // return 4


Algorithm:

-I will use an OrderedDict to store the keys and values while maintaining the order of insertion.
- get(key):
    -If the key is in the OrderedDict, move the key to the end (most recently used) and return its value.
    -If the key is not in the OrderedDict, return -1.
- put(key, value):
    -If the key is in the OrderedDict, update its value and move it to the end.
    -If the key is not in the OrderedDict and the capacity is full, remove the first (least recently used) key-value pair.
    -Add the key-value pair to the OrderedDict.
    
Time Complexity:

O(1) for both get and put operations. This is because both operations are implemented in constant time using OrderedDict.

Space Complexity:
O(capacity), where capacity is the maximum number of elements the cache can hold.

           [[], [1], [2], [1], [3], [2], [2], [2], [2]]

    - Explanation

        - myHashSet.add(2);      // set = [1, 2]

        - myHashSet.contains(3); // return False, (not found)
        - myHashSet.add(2);      // set = [1, 2]
        - myHashSet.contains(2); // return True
        - myHashSet.remove(2);   // set = [1]
        - myHashSet.contains(2); // return False, (already removed)
 

Algorithm:
    
- Use a list of lists (buckets) to handle hash collisions.
- Use the modulo operator to determine the bucket index for a given key.
- Implement add, contains, and remove operations by performing operations within the appropriate bucket.

This HashSet works by using a list of boolean values. When an element is added to the set, the value at the index corresponding to that element is set to True. When an element is removed, the value at that index is set to False. To check if an element is in the set, we simply return the value at the index corresponding to that element.

Time Complexity
O(1) on average for add, contains, and remove operations assuming the hash function distributes keys uniformly.
Worst-case time complexity is O(n) if all keys hash to the same bucket.
Space Complexity
O(n), where n is the number of elements in the HashSet.

**Problem5-Leetcode#706- Design HashMap-Easy**

Design a HashMap without using any built-in hash table libraries.

Implement the MyHashMap class:

MyHashMap() initializes the object with an empty map.\
void put(int key, int value) inserts a (key, value) pair into the HashMap.\
If the key already exists in the map, update the corresponding value.
int get(int key) returns the value to which the specified key is mapped,\
or -1 if this map contains no mapping for the key.\
void remove(key) removes the key and its corresponding value\
if the map contains the mapping for the key.
 

- Example 1:

    - Input ["MyHashMap", "put", "put", "get", "get", "put", "get", "remove", "get"]\
          [[], [1, 1], [2, 2], [1], [3], [2, 1], [2], [2], [2]]
    - Output [null, null, null, 1, -1, null, 1, null, -1]

    - Explanation
        - MyHashMap myHashMap = new MyHashMap();
        - myHashMap.put(1, 1); // The map is now [[1,1]]
        - myHashMap.put(2, 2); // The map is now [[1,1], [2,2]]
        - myHashMap.get(1);    // return 1, The map is now [[1,1], [2,2]]
        - myHashMap.get(3);    // return -1 (i.e., not found), The map is now [[1,1], [2,2]]
        - myHashMap.put(2, 1); // The map is now [[1,1], [2,1]] (i.e., update the existing value)
        - myHashMap.get(2);    // return 1, The map is now [[1,1], [2,1]]
        - myHashMap.remove(2); // remove the mapping for 2, The map is now [[1,1]]
        - myHashMap.get(2);    // return -1 (i.e., not found), The map is now [[1,1]]


This algorithm is for designing a simple hashmap, which is a type of data structure that can store key-value pairs for quick lookup.
The approach I will follow for this-
- The algorithm uses a list of lists (or “buckets”) to store the key-value pairs. Each bucket can hold one or more pairs. This is a common way to handle “hash collisions”, which happen when two different keys have the same hash value.
- To decide which bucket to put a key-value pair in, the algorithm uses the modulo operator. It takes the hash of the key and divides it by the number of buckets. The remainder is the index of the bucket where the pair should be stored. This ensures that the pairs are spread out evenly across the buckets.
- To add a new key-value pair, the algorithm calculates the bucket index for the key and adds the pair to the appropriate bucket.
- To retrieve the value for a key, the algorithm calculates the bucket index for the key, then searches through that bucket to find the pair with the matching key.
- To remove a key-value pair, the algorithm calculates the bucket index for the key, then searches through that bucket to find and remove the pair with the matching key.

Time Complexity:
O(1) on average for put, get, and remove operations assuming the hash function distributes keys uniformly.
Worst-case time complexity is O(n) if all keys hash to the same bucket.

Space Complexity:
O(n), where n is the number of elements in the HashMap.
