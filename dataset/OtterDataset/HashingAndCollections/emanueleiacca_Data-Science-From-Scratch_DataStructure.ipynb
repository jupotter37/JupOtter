##### üîó **1. Singly Linked List**

A **Singly Linked List (SLL)** is a data structure where each node points to the **next node** in the sequence, and the last node points to `None`.

 **Mathematical Representation**

Each node contains:
- A **data field** that stores the information.
- A **next pointer** that points to the next node.

$$
\text{Node}_1 \rightarrow \text{Node}_2 \rightarrow \text{Node}_3 \rightarrow \cdots \rightarrow \text{Node}_n \rightarrow \textbf{None}
$$

‚úÖ **Advantages**
- Dynamic size.
- Efficient memory usage.

‚ùå **Disadvantages**
- Cannot traverse backward.
- Searching takes \( O(n) \) time.


message delivery on network (message is broken into packets and each packet has a key of the next one so that at the receiver's end , it will be easy to club them)


```python
# Node class for Singly Linked List
class IrisNode:
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width, species):
        """Initialize the node with the given values."""
        self.sepal_length = sepal_length 
        self.sepal_width = sepal_width
        self.petal_length = petal_length
        self.petal_width = petal_width
        self.species = species
        self.next = None

# Singly Linked List class
class IrisSinglyLinkedList:
    """Singly Linked List to store Iris flowers."""
    def __init__(self):
        self.head = None # Initialize the head node to None
    
    def add_flower(self, sepal_length, sepal_width, petal_length, petal_width, species):
        """Add a new flower to the beginning of the linked list."""
        new_node = IrisNode(sepal_length, sepal_width, petal_length, petal_width, species) # Create a new node
        new_node.next = self.head # Set the next of the new node to the current head
        self.head = new_node 
    
    def print_flowers(self, limit=5):
        """Print the flowers in the linked list."""
        current = self.head
        count = 0
        while current and count < limit: # Iterate through the linked list
            print(f"Sepal Length: {current.sepal_length}, Sepal Width: {current.sepal_width}, "
                  f"Petal Length: {current.petal_length}, Petal Width: {current.petal_width}, "
                  f"Species: {current.species}")
            current = current.next # Move to the next node
            count += 1

iris_linked_list = IrisSinglyLinkedList()

for index, row in iris_df.head(5).iterrows():
    iris_linked_list.add_flower(row['sepal length (cm)'], row['sepal width (cm)'],
                                row['petal length (cm)'], row['petal width (cm)'], row['species'])

print("First 5 Flowers in Singly Linked List:")
iris_linked_list.print_flowers()
```

##### üîÅ **2. Doubly Linked List**

A **Doubly Linked List (DLL)** is a type of linked list where each node has **two pointers**:
- One pointing to the **next node**.
- One pointing to the **previous node**.

üìä **Mathematical Representation**

$$
\textbf{None} \leftarrow \text{Node}_1 \leftrightarrow \text{Node}_2 \leftrightarrow \cdots \leftrightarrow \text{Node}_n \rightarrow \textbf{None}
$$

‚úÖ **Advantages**
- Bi-directional traversal.
- Easier node deletion.

‚ùå **Disadvantages**
- Extra memory overhead for `prev` pointers.
- Slightly complex implementation.

DNA molecules


```python
# Node class for Doubly Linked List
class IrisDoublyNode:
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width, species):
        self.sepal_length = sepal_length
        self.sepal_width = sepal_width
        self.petal_length = petal_length
        self.petal_width = petal_width
        self.species = species
        self.next = None
        self.prev = None # Add a new pointer to the previous node

# Doubly Linked List class
class IrisDoublyLinkedList:
    def __init__(self):
        # Initialize head and tail to None
        self.head = None 
        self.tail = None
    
    def add_flower(self, sepal_length, sepal_width, petal_length, petal_width, species):
        new_node = IrisDoublyNode(sepal_length, sepal_width, petal_length, petal_width, species)
        if not self.head: # If the list is empty
            self.head = self.tail = new_node # set the head and tail to the new node
        else: 
            self.tail.next = new_node # Set the next of the current tail to the new node
            new_node.prev = self.tail # Set the previous of the new node to the current tail
            self.tail = new_node # Set the tail to the new node
    
    def traverse_forward(self, limit=5):
        """Traverse the linked list in the forward direction."""
        current = self.head # Start from the head
        count = 0
        print("‚û°Ô∏è Forward Traversal:")
        while current and count < limit: # Iterate through the linked list
            print(f"{current.species}: Sepal({current.sepal_length}, {current.sepal_width}), "
                  f"Petal({current.petal_length}, {current.petal_width})")
            current = current.next # Move to the next node
            count += 1
    
    def traverse_backward(self, limit=5):
        """Traverse the linked list in the backward direction"""
        current = self.tail # Start from the tail
        count = 0
        print("‚¨ÖÔ∏è Backward Traversal:")
        while current and count < limit: 
            print(f"{current.species}: Sepal({current.sepal_length}, {current.sepal_width}), "
                  f"Petal({current.petal_length}, {current.petal_width})")
            current = current.prev # Move to the previous node
            count += 1

iris_doubly_list = IrisDoublyLinkedList()

for index, row in iris_df.head(5).iterrows():
    iris_doubly_list.add_flower(row['sepal length (cm)'], row['sepal width (cm)'],
                                row['petal length (cm)'], row['petal width (cm)'], row['species'])

iris_doubly_list.traverse_forward()

iris_doubly_list.traverse_backward()
```

##### üîÑ **3. Circular Linked List**

A **Circular Linked List (CLL)** is a variation where:
- The **last node points back to the first node**, forming a circular structure.

üìä **Mathematical Representation**

**Singly Circular Linked List:**
$$
\text{Node}_1 \rightarrow \text{Node}_2 \rightarrow \cdots \rightarrow \text{Node}_n \rightarrow \text{Node}_1
$$

**Doubly Circular Linked List:**
$$
\text{Node}_1 \leftrightarrow \text{Node}_2 \leftrightarrow \cdots \leftrightarrow \text{Node}_n \leftrightarrow \text{Node}_1
$$

 ‚úÖ **Advantages**
- Infinite traversal capability.
- Optimized for circular or repeating tasks.

‚ùå **Disadvantages**
- Slightly complex traversal logic.
- Debugging errors may cause infinite loops.

Multiple Player board game


```python
# Circular Linked List Node
class IrisCircularNode:
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width, species):
        self.sepal_length = sepal_length
        self.sepal_width = sepal_width
        self.petal_length = petal_length
        self.petal_width = petal_width
        self.species = species
        self.next = None

# Circular Linked List
class IrisCircularLinkedList:
    def __init__(self):
        self.head = None

    def add_flower(self, sepal_length, sepal_width, petal_length, petal_width, species):
        """Add a flower node to the circular linked list."""
        new_node = IrisCircularNode(sepal_length, sepal_width, petal_length, petal_width, species) # Create a new node
        if not self.head:
            self.head = new_node 
            new_node.next = self.head  # Point to itself to make it circular
        else:
            current = self.head # Start from the head
            while current.next != self.head: # Traverse to the last node
                current = current.next # Move to the next node
            current.next = new_node # Set the next of the last node to the new node
            new_node.next = self.head # Set the next of the new node to the head

    def print_species(self, limit=10):
        """Print the circular linked list with a limit on the number of cycles."""
        if not self.head:
            print("The circular linked list is empty.")
            return

        current = self.head # Start from the head
        count = 0
        print("üîÑ Circular Linked List Traversal (Limited to", limit, "cycles):")
        while count < limit: # Iterate through the linked list
            print(
                f"{current.species}: Sepal({current.sepal_length}, {current.sepal_width}), "
                f"Petal({current.petal_length}, {current.petal_width})"
            )
            current = current.next # Move to the next node
            count += 1
            if current == self.head: # If we reach the head node again
                print("‚Ü©Ô∏è Returned to Head Node (Restarting Cycle)")

iris_circular_list = IrisCircularLinkedList()

for index, row in iris_df.head(5).iterrows():
    iris_circular_list.add_flower(
                    row['sepal length (cm)'], row['sepal width (cm)'],
                    row['petal length (cm)'], row['petal width (cm)'], row['species']
    )

# Print species in a circular fashion
iris_circular_list.print_species(limit=10)
```

### üìö **Hashing and Universal Hashing Overview**
‚úÖ Pros:
- Fast lookups and insertions.

**Hashing**
- Hashing efficiently stores and retrieves data using a **hash table**.
- A **hash function** maps data items to **positions (buckets)** in the table.
- **Advantages:**  
   - Reduced memory usage.  
   - Faster lookups compared to linked lists.
- A **hash function (h(x))** maps data to a finite set of buckets:
  $$
  h(x): \text{Data Item} \rightarrow \text{Bucket Index}
  $$
- **Goal:** Evenly distribute data across buckets to minimize collisions.  
- Poor hash functions can lead to **clustering**, reducing efficiency.



```python
# Basic Hash Table with Modulo Hashing
class BasicHashTable:
    """Basic hash table with modulo hashing"""
    def __init__(self, size):
        """Initialize hash table"""
        self.size = size # Number of buckets
        self.table = [None] * size # Initialize hash table with empty buckets

    def hash_function(self, key):
        """Hash function to determine index of key-value pair"""
        return int(key) % self.size  # Modulo hashing

    def insert(self, key, value):
        """Insert key-value pair into hash table"""
        index = self.hash_function(key) # Get index from hash function
        if self.table[index] is None: # If bucket is empty:
            self.table[index] = [] # Create new list
        self.table[index].append((key, value)) # Append key-value pair to bucket

    def retrieve(self, key):
        """Retrieve value from hash table using key"""
        index = self.hash_function(key) # Get index from hash function
        if self.table[index]: # If bucket is not empty:
            for k, v in self.table[index]: # Iterate through key-value pairs in bucket
                if k == key: # If key matches:
                    return v 
        return None

basic_ht = BasicHashTable(size=10)

for key, value in dataset:
    basic_ht.insert(key, value) # Insert key-value pair into hash table

print("\nRetrieve from Basic Hash Table:")
for i in range(5):
    print(f"Key: {dataset[i][0]} -> Value: {basic_ht.retrieve(dataset[i][0])}")

# Collisions are handled using chaining (linked list in buckets).
```

 **Universal Hashing**
- **Problem:** No single hash function works optimally for all datasets.
- **Solution:** Use a **family of hash functions (H)** and randomly select one.
- **Definition:** A hash family is *universal* if:
  $$
  P(h(x) = h(y)) = \frac{1}{n}, \quad \text{for all } x \neq y
  $$
  - *n:* Number of buckets.


- From number theory, for a hash table with **n prime buckets**:
  $$
  h_a(x) = (a_1 x_1 + a_2 x_2 + \ldots + a_k x_k) \mod n
  $$
  - *x:* Data represented as a k-tuple.
  - *a:* Randomly selected coefficients from \(\{0, ‚Ä¶, n-1\}\).
  - *n:* Number of buckets.

- Minimizes **collisions** with high probability.
- Ensures **good average-case performance**.
- Combines **randomization** with **mathematical guarantees**.


```python
import random
# Universal Hash Table
class UniversalHashTable:
    def __init__(self, size, prime):
        self.size = size # Number of buckets
        self.prime = prime # Prime number for modulo hashing
        self.a = random.randint(1, prime - 1)  # Random coefficient
        self.b = random.randint(0, prime - 1)  # Random offset
        self.table = [None] * size # Initialize hash table with empty buckets

    def hash_function(self, key):
        """The hash function avoids clustering by using a randomized formula"""
        return ((self.a * key + self.b) % self.prime) % self.size # Universal hashing

    def insert(self, key, value): # same as before
        index = self.hash_function(key)
        if self.table[index] is None:
            self.table[index] = []
        self.table[index].append((key, value))

    def retrieve(self, key): # same as before
        index = self.hash_function(key)
        if self.table[index]:
            for k, v in self.table[index]:
                if k == key:
                    return v
        return None

universal_ht = UniversalHashTable(size=10, prime=31)

for key, value in dataset:
    universal_ht.insert(key, value)

print("\nRetrieve from Universal Hash Table:")
for i in range(5):
    print(f"Key: {dataset[i][0]} -> Value: {universal_ht.retrieve(dataset[i][0])}")
```

```python
import matplotlib.pyplot as plt

def plot_hash_distribution(ht, dataset, title):
    hash_indices = [ht.hash_function(key) for key, _ in dataset]
    plt.hist(hash_indices, bins=ht.size, edgecolor='black')
    plt.title(title)
    plt.xlabel('Hash Bucket')
    plt.ylabel('Number of Keys')
    plt.show()

# Plot Distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plot_hash_distribution(basic_ht, dataset, "Basic Hashing Distribution")

plt.subplot(1, 2, 2)
plot_hash_distribution(universal_ht, dataset, "Universal Hashing Distribution")
```

1. Basic Hashing may not account for skewed data distribution, leading to potential clustering.

2. Universal Hashing aims to reduce clustering by introducing randomness into bucket assignment.

By selecting a hash function at random from a universal family, the probability of collisions is minimized.
This ensures better average-case performance across diverse datasets.

```python
import time

def benchmark_hash_table(ht, dataset):
    start_time = time.time()
    for key, _ in dataset:
        ht.retrieve(key)
    end_time = time.time()
    return end_time - start_time

basic_time = benchmark_hash_table(basic_ht, dataset)
universal_time = benchmark_hash_table(universal_ht, dataset)

print("\nPerformance Comparison:")
print(f"Basic Hash Table Lookup Time: {basic_time:.4f} seconds")
print(f"Universal Hash Table Lookup Time: {universal_time:.4f} seconds")
```

## Abstract Data Structures

### Stack

A LIFO (Last In, First Out) data structure: elements can be inserted
one after the other, and the last element inserted is the first element to be extracted.

A stack S supports the following operations:
-  S.push(x): Insert element x on the top of Q.
-  S.pop(): Remove element form the top of Q and return it.
-  S.peek(): Return the element on the top of Q.
-  S.isEmpty() : Return True if Q is empty.

Use Case: Undo/Redo, function call stack.

```python
# Stack Implementation (LIFO Paradigm)

class Stack:
    def __init__(self):
        self.stack = []  # Use a list to store stack items
    
    def push(self, x):
        """Insert element x on the top of the stack."""
        self.stack.append(x)  # Add item at the end of the list (top of the stack)
    
    def pop(self):
        """Remove and return the top element of the stack."""
        if self.isEmpty():
            return None
        return self.stack.pop()  # Remove and return the last element
    
    def peek(self):
        """Return the top element without removing it."""
        if self.isEmpty():
            return None
        return self.stack[-1]  # Access the last element
    
    def isEmpty(self):
        """Check if the stack is empty."""
        return len(self.stack) == 0

# Group species by average petal length
species_avg_petal_length = iris_df.groupby('species')['petal length (cm)'].mean().to_dict()

# Create a Stack
iris_stack = Stack()

# Push species with their average petal length onto the stack
for species, avg_petal_length in species_avg_petal_length.items():
    iris_stack.push((species, avg_petal_length))

# Test Operations
print("üîÑ Testing Stack Operations\n")

# Peek at the top
print("üëÄ Peek Top:", iris_stack.peek())

# Pop elements one by one
while not iris_stack.isEmpty():
    top_item = iris_stack.pop()
    print(f"üì¶ Popped -> Species: {top_item[0]}, Average Petal Length: {top_item[1]:.2f}")
```

‚úÖ Pros:

- Efficient push and pop operations.

‚ùå Cons:
- Limited access (only the top element).

### Map

The map ADT is a data structure that allows to associate a key with a value. 

A map M supports the following operations.
- M.set(k, v): Set the value of k in M equal to k. Replace the existing value associated with k if k
already exists in M .
- M.get(k): Return the value associated with key k in M .
- M.exists(k): Return True if k exists in M , False if it does not.
- M.delete(k): Remove the key k from M and its associated value.
- M.isEmpty() : Return True if M is empty.



An example of a stack S. Elements are inserted on the top of S and removed from the top.
- S.pop() will remove 1 from S and return it. 
- The sequence S.push(9); S.push(2), will insert 9 on top of 1
and 2 on top of 9, and will not return any value. 
- Therefore, the sequence S.push(9); S.peek(); S.push(2);
S.pop(); S.pop(); S.peek(); S.pop(); S.pop(), will return the values 9, 2, 9, 1, 1, and 3.



```python
# Map Implementation (Key-Value Store)

class Map:
    def __init__(self):
        self.map = {}  # Use a dictionary to store key-value pairs
    
    def set(self, k, v):
        """Set the value associated with key k."""
        self.map[k] = v  # Add or update the key-value pair
    
    def get(self, k):
        """Return the value associated with key k."""
        return self.map.get(k, None)  # Return the value or None if key does not exist
    
    def exists(self, k):
        """Check if the key k exists in the map."""
        return k in self.map  # Return True if key exists, False otherwise
    
    def delete(self, k):
        """Remove the key k and its associated value from the map."""
        if k in self.map:
            del self.map[k]  # Remove the key-value pair
    
    def isEmpty(self):
        """Check if the map is empty."""
        return len(self.map) == 0

# Group species by average petal length
species_avg_petal_length = iris_df.groupby('species')['petal length (cm)'].mean().to_dict()

# Create a Map
iris_map = Map()

# Set species with their average petal length
for species, avg_petal_length in species_avg_petal_length.items():
    iris_map.set(species, avg_petal_length)

# Test Operations
print("üîÑ Testing Map Operations\n")

# Check if a specific species exists
species_to_check = 'setosa'
print(f"üîç Exists '{species_to_check}':", iris_map.exists(species_to_check))

# Get a species' average petal length
print(f"üåº Average Petal Length of '{species_to_check}':", iris_map.get(species_to_check))

# Delete a species
species_to_delete = 'versicolor'
iris_map.delete(species_to_delete)
print(f"‚ùå Deleted '{species_to_delete}'. Exists now?", iris_map.exists(species_to_delete))

# Check if the map is empty
print("ü´† Is Map Empty?", iris_map.isEmpty())
```

### Queue

A FIFO (First In, First Out) data structure: elements can be inserted
to Q one after the other, and the first one to be inserted is the last one to be extracted

A queue Q supports the following operations.
- Q.enqueue(x): Insert element x to the back of Q.
- Q.dequeue(): Remove the element in the front of Q and return it.
- Q.peek(): Return the elements in the front of Q.
- Q.isEmpty() : Return True if Q is empty.



An example of a queue Q. Elements are inserted to the back of Q and returned from the front.

- Q.peek() will return 3, and leave Q as is. Q.dequeue() will return 3 and remove it, so that the next element in
the front of the queue will be 1. 

- Therefore, the sequence Q.dequeue(); Q.peek(); Q.dequeue(); Q.dequeue(),
will return the values 3, 1, 1, and 2. 

- The sequence of Q.enqueue(7); Q.enqueue(9), will put 7 to the back,
after 6, and then 9 after 7.

Use Case: Job scheduling, printer queues.

```python
# Queue Implementation (FIFO Paradigm)

class Queue:
    def __init__(self):
        self.q = []  # Use a list to store queue items
    
    def enqueue(self, x):
        """Insert element x to the back of the queue."""
        self.q.append(x)  # Add item at the end of the list
    
    def dequeue(self):
        """Remove and return the front element of the queue."""
        if self.isEmpty():
            return None
        return self.q.pop(0)  # Remove and return the first element
    
    def peek(self):
        """Return the front element without removing it."""
        if self.isEmpty():
            return None
        return self.q[0]  # Access the first element
    
    def isEmpty(self):
        """Check if the queue is empty."""
        return len(self.q) == 0

# Group species by average petal length
species_avg_petal_length = iris_df.groupby('species')['petal length (cm)'].mean().to_dict()

# Create a Queue
iris_queue = Queue()

# Enqueue species with their average petal length
for species, avg_petal_length in species_avg_petal_length.items():
    iris_queue.enqueue((species, avg_petal_length))

# Test Operations
print("üîÑ Testing Queue Operations\n")

# Peek at the front
print("üëÄ Peek Front:", iris_queue.peek())

# Dequeue elements one by one, from smallest to largest
while not iris_queue.isEmpty():
    front_item = iris_queue.dequeue()
    print(f"üöÄ Dequeued -> Species: {front_item[0]}, Average Petal Length: {front_item[1]:.2f}")
```

‚úÖ Pros:
- Efficient for scheduling tasks.

‚ùå Cons:
- Limited direct access to elements.

### üìö **Heap Data Structure and Related Concepts**

**Binary Heap Representation:**  
- Typically implemented using an **array**.
- For a node at index `i`:
   - **Left child:** `2 * i + 1`
   - **Right child:** `2 * i + 2`
   - **Parent:** `(i - 1) // 2`

### **1. Max Heap**
- A **Max Heap** ensures the **largest element** is always at the **root**.
- **Properties:**
   - Parent nodes are always **greater than or equal** to their child nodes.
   - Root node contains the **maximum value**.

**Example:**

         50
        /  \
      30  40
     / \  / \
    10 20 35 25


**Operations:**
- **Insert:** Add a new element and adjust to maintain the heap property.
- **Delete (Max Extract):** Remove the root and adjust the heap.


**Best Case Scenario:**
Example: Adding an element smaller than all other elements (No bubbling up)
- **Insertion:** O(1) ‚Äì The new element is inserted directly without requiring reordering.
- **Extraction (Max Element):** O(1) ‚Äì The root is accessed directly, and no reordering is required.

**Worst Case Scenario:**
Example: Adding an element larger than the root (Bubbles up to root)
- **Insertion:** O(log n) ‚Äì The new element has to be "bubbled up" to the correct position.
- **Extraction (Max Element):** O(log n) ‚Äì The heap must reorder nodes after removing the root.


```python
class MaxHeap:
    def __init__(self):
        self.heap = [] # Initialize an empty list to store the heap elements
    
    def parent(self, i):
        return (i - 1) // 2 # Return the parent index of the given index
    
    def left_child(self, i):
        return 2 * i + 1 # Return the left child index of the given index
    
    def right_child(self, i):
        return 2 * i + 2 # Return the right child index of the given index
    
    def insert(self, value):
        self.heap.append(value) # Append the value to the heap
        self._heapify_up(len(self.heap) - 1) # Heapify up from the last index
    def _heapify_up(self, index):
        """
        insert a new element to a heap ( we add it at the bottom of the heap tree, and move up the tree while comparing to the current parent element and swapping if needed)
        """
        while index > 0 and self.heap[self.parent(index)] < self.heap[index]: # comparing to the current parent element
            self.heap[self.parent(index)], self.heap[index] = self.heap[index], self.heap[self.parent(index)] # Swap with parent if greater
            index = self.parent(index) # Move up the tree
    
    def extract_max(self):
        if len(self.heap) == 0: # If heap is empty
            return None
        root = self.heap[0] # Get the root of the heap
        self.heap[0] = self.heap.pop() # Remove the last element and set it as the root
        self._heapify_down(0) # Heapify down from the root
        return root
    
    def _heapify_down(self, index):
        """ heapify down the tree, we compare the current element with its children and swap with the largest child if needed"""
        largest = index # Initialize largest as the root
        left = self.left_child(index) # Get the left child
        right = self.right_child(index) # Get the right child
        
        if left < len(self.heap) and self.heap[left] > self.heap[largest]: # Compare with left child
            largest = left # Set left as largest if greater
        if right < len(self.heap) and self.heap[right] > self.heap[largest]: # Compare with right child
            largest = right # Set right as largest if greater
        if largest != index: # If the largest element is not the root
            self.heap[index], self.heap[largest] = self.heap[largest], self.heap[index] # Swap with the largest
            self._heapify_down(largest) # Recursively heapify down

max_heap = MaxHeap()
max_heap.insert(10)
max_heap.insert(20)
max_heap.insert(15)
print("Max Heap Array:", max_heap.heap)
print("Extract Max:", max_heap.extract_max())
print("Max Heap after Extract:", max_heap.heap)
```

###  **2. Min Heap**
- A **Min Heap** ensures the **smallest element** is always at the **root**.
- **Properties:**
   - Parent nodes are always **smaller than or equal** to their child nodes.
   - Root node contains the **minimum value**.

**Example:**

         10
        /  \
      20  30
     / \  / \
    40 50 60 70

    
**Operations:**
- **Insert:** Add a new element and adjust to maintain the heap property.
- **Delete (Min Extract):** Remove the root and adjust the heap.

**Best Case Scenario:**
Example: Adding an element larger than all other elements.  
- **Insertion:** O(1) ‚Äì The new element is already in the correct position at the leaf.
- **Extraction (Min Element):** O(1) ‚Äì Directly access the root without adjustment.

**Worst Case Scenario:**
Example: Adding an element smaller than the root (Bubbles up to root).
- **Insertion:** O(log n) ‚Äì The new element moves up to the root (bubbles up).
- **Extraction (Min Element):** O(log n) ‚Äì After extracting the root, the last element must bubble down to the correct spot.

```python
class MinHeap:
    # All the same
    def __init__(self):
        self.heap = []
    
    def parent(self, i):
        return (i - 1) // 2
    
    def left_child(self, i):
        return 2 * i + 1
    
    def right_child(self, i):
        return 2 * i + 2
    
    def insert(self, value):
        self.heap.append(value)
        self._heapify_up(len(self.heap) - 1)
    
    def _heapify_up(self, index):
        while index > 0 and self.heap[self.parent(index)] > self.heap[index]:
            self.heap[self.parent(index)], self.heap[index] = self.heap[index], self.heap[self.parent(index)]
            index = self.parent(index)
    def extract_min(self):
        if len(self.heap) == 0:
            return None
        root = self.heap[0]
        self.heap[0] = self.heap.pop()
        self._heapify_down(0)
        return root
    
    # Until here  
    def _heapify_down(self, index):
        smallest = index # we look for smallest instead of largest
        left = self.left_child(index)
        right = self.right_child(index)
        
        if left < len(self.heap) and self.heap[left] < self.heap[smallest]: # so hear the sign is minor
            smallest = left
        if right < len(self.heap) and self.heap[right] < self.heap[smallest]: # also here
            smallest = right
        if smallest != index:
            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            self._heapify_down(smallest)

# Example
min_heap = MinHeap()
min_heap.insert(10)
min_heap.insert(20)
min_heap.insert(5)
print("Min Heap Array:", min_heap.heap)
print("Extract Min:", min_heap.extract_min())
print("Min Heap after Extract:", min_heap.heap)
```

### Binary Heap
supports both Min-Heap and Max-Heap modes.

**Best Case Scenario:**
Example: Balanced binary heap built from a sorted array.  
- **Build Heap:** O(n) ‚Äì Linear time when using the bottom-up approach.
- **Search:** O(1) ‚Äì Accessing the root element is always O(1).

**Worst Case Scenario:**
Example: Skewed binary heap requiring multiple swaps during insertion.
- **Build Heap:** O(n log n) ‚Äì When elements are inserted one by one (top-down approach).
- **Search:** O(n) ‚Äì Finding an arbitrary element in the heap.

```python
class BinaryHeap:
    def __init__(self, heap_type='min'):
        """
        Initialize the binary heap.
        heap_type: 'min' for Min-Heap, 'max' for Max-Heap.
        """
        self.heap = []
        self.heap_type = heap_type

    def parent(self, i):
        """Return the index of the parent node."""
        return (i - 1) // 2

    def left_child(self, i):
        """Return the index of the left child node."""
        return 2 * i + 1

    def right_child(self, i):
        """Return the index of the right child node."""
        return 2 * i + 2

    def compare(self, child, parent):
        """
        Comparison function for min-heap or max-heap.
        """
        if self.heap_type == 'min':
            return self.heap[child] < self.heap[parent]
        else:
            return self.heap[child] > self.heap[parent]

    def insert(self, value):
        """
        Insert a value into the heap.
        """
        self.heap.append(value)
        self._heapify_up(len(self.heap) - 1)

    def _heapify_up(self, index):
        """
        Restore the heap property upwards.
        """
        while index > 0 and self.compare(index, self.parent(index)):
            self.heap[index], self.heap[self.parent(index)] = (
                self.heap[self.parent(index)],
                self.heap[index],
            )
            index = self.parent(index)

    def extract_root(self):
        """
        Extract the root element (min or max).
        """
        if len(self.heap) == 0:
            return None
        root = self.heap[0]
        self.heap[0] = self.heap.pop()
        self._heapify_down(0)
        return root

    def _heapify_down(self, index):
        """
        Restore the heap property downwards.
        """
        size = len(self.heap)
        while True:
            smallest_or_largest = index
            left = self.left_child(index)
            right = self.right_child(index)

            if left < size and self.compare(left, smallest_or_largest):
                smallest_or_largest = left
            if right < size and self.compare(right, smallest_or_largest):
                smallest_or_largest = right

            if smallest_or_largest != index:
                self.heap[index], self.heap[smallest_or_largest] = (
                    self.heap[smallest_or_largest],
                    self.heap[index],
                )
                index = smallest_or_largest
            else:
                break

    def peek(self):
        """
        Get the root value without removing it.
        """
        return self.heap[0] if self.heap else None

    def is_empty(self):
        """
        Check if the heap is empty.
        """
        return len(self.heap) == 0

    def display(self):
        """
        Display the heap.
        """
        print(self.heap)
```

```python
# Min Heap Example
print("‚úÖ Min Heap Example")
min_heap = BinaryHeap(heap_type='min')
min_heap.insert(10)
min_heap.insert(4)
min_heap.insert(15)
min_heap.insert(20)
min_heap.insert(1)

print("Heap Array (Min Heap):", min_heap.heap)
print("Root (Min Element):", min_heap.peek())
print("Extract Root:", min_heap.extract_root())
print("Heap after extraction:", min_heap.heap)
```

###  **3. Heap Sort**
- **Heap Sort** is a **comparison-based sorting algorithm** using the heap data structure.
- It can be implemented using either a **Max Heap** or **Min Heap**.

**Algorithm Steps (Max Heap):**
1. Build a **Max Heap** from the input array.
2. Swap the **root node** (maximum value) with the **last node**.
3. Remove the last node and **heapify** the root node.
4. Repeat until the heap is empty.

 **Time Complexity:**
- **Build Heap:** $ O(n) $
- **Heapify:** $ O(\log n) $
- **Overall Complexity:** $ O(n \log n) $

**Advantages:**
- In-place algorithm (no extra space required).
- Consistent time complexity.

**Disadvantages:**
- Not stable (relative order of equal elements is not preserved).


**Best Case Scenario:**
- **Time Complexity:** O(n log n) ‚Äì Consistent across all scenarios because the structure requires heapify at each step.

**Worst Case Scenario:**

- **Time Complexity:** O(n log n) ‚Äì No degradation in time complexity

```python
class HeapSort:
    def __init__(self, heap_type='min'):
        self.heap_type = heap_type  # Define if it's a min-heap or max-heap

    def heapify(self, arr, n, i):
        """
        Maintain the heap property for the given subtree rooted at index i.
        """
        largest = i  # Initialize root as largest for max-heap
        smallest = i  # Initialize root as smallest for min-heap
        left = 2 * i + 1  # Get the left child index
        right = 2 * i + 2  # Get the right child index

        if self.heap_type == 'max':
            # For Max-Heap: Find the largest among root, left, and right child
            if left < n and arr[left] > arr[largest]: 
                largest = left
            if right < n and arr[right] > arr[largest]:
                largest = right

            # Swap and continue heapifying if root is not the largest
            if largest != i:
                arr[i], arr[largest] = arr[largest], arr[i]
                self.heapify(arr, n, largest)

        elif self.heap_type == 'min':
            # For Min-Heap: Find the smallest among root, left, and right child
            if left < n and arr[left] < arr[smallest]:
                smallest = left
            if right < n and arr[right] < arr[smallest]:
                smallest = right

            # Swap and continue heapifying if root is not the smallest
            if smallest != i:
                arr[i], arr[smallest] = arr[smallest], arr[i]
                self.heapify(arr, n, smallest)

    def heap_sort(self, arr):
        """
        Perform heap sort on the given array.
        """
        n = len(arr)

        # Build heap (rearrange array into heap)
        for i in range(n // 2 - 1, -1, -1):
            self.heapify(arr, n, i)

        # Extract elements one by one from the heap
        for i in range(n - 1, 0, -1):
            arr[i], arr[0] = arr[0], arr[i]  # Swap the root with the last element
            self.heapify(arr, i, 0)  # Heapify the reduced heap

arr = [4, 10, 3, 5, 1]
print("Original Array:", arr)

# Max-Heap Sort
heap_sort = HeapSort(heap_type='max')
heap_sort.heap_sort(arr)
print("Sorted Array (Max-Heap):", arr)

# Min-Heap Sort
arr = [4, 10, 3, 5, 1]
heap_sort = HeapSort(heap_type='min')
heap_sort.heap_sort(arr)
print("Sorted Array (Min-Heap):", arr)
```

### Priority Queue

A Priority Queue is a special type of queue where elements are dequeued based on their priority, not their order of arrival. It allows to insert elements each of which has a priority, and it allows to extract the
one with the maximum priority


- Q.insert(x): Insert element x to the back of Q.
- Q.removeMax(): Remove the maximum element of Q and return it.
- Q.peekMax(): Return the maximum elements in Q.
- Q.isEmpty() : Return True if Q is empty.

```python
import heapq

species_avg_petal_length = iris_df.groupby('species')['petal length (cm)'].mean().to_dict()

# Priority Queue with Max Operations
class MaxPriorityQueue:
    def __init__(self):
        self.pq = []  # Use a list to store heap items
    
    def insert(self, x):
        # Use negative priority for max heap simulation with heapq
        heapq.heappush(self.pq, (-x[0], x[1])) # Insert as a tuple (-priority, item), negative because heapq is a min heap
    
    def removeMax(self):
        if self.isEmpty():
            return None
        max_item = heapq.heappop(self.pq) # Pop the maximum item
        return (-max_item[0], max_item[1])  
    
    def peekMax(self):
        if self.isEmpty():
            return None
        return (-self.pq[0][0], self.pq[0][1]) # Return the maximum item
    
    def isEmpty(self):
        return len(self.pq) == 0

# Create a Max Priority Queue
iris_max_pq = MaxPriorityQueue()

# Insert species with their average petal length
for species, avg_petal_length in species_avg_petal_length.items():
    iris_max_pq.insert((avg_petal_length, species))

# Test Operations
print("üîÑ Testing Priority Queue Operations\n")

# Peek at the maximum
print("üëÄ Peek Max:", iris_max_pq.peekMax())

# Remove max elements one by one
while not iris_max_pq.isEmpty():
    max_item = iris_max_pq.removeMax()
    print(f"üåü Removed Max -> Species: {max_item[1]}, Average Petal Length: {max_item[0]:.2f}")


```

### Tree
A tree is a hierarchical data structure consisting of nodes connected by edges. A binary tree ensures every node has at most two children.

- Use Case: File systems, database indexes

```python
class IrisNode:
    def __init__(self, feature, value):
        """Initialize the node with the given feature and value."""
        self.feature = feature # Feature name
        self.value = value # Feature value
        self.left = None # Pointer to the left child
        self.right = None # Pointer to the right child

class IrisTree:
    def __init__(self):
        self.root = None # Initialize the root node to None

    def insert(self, feature, value):
        """Insert a new node into the tree."""
        if not self.root: # If the tree is empty
            self.root = IrisNode(feature, value) # Set the root to the new node
        else:
            self._insert_recursive(self.root, feature, value) # Insert recursively

    def _insert_recursive(self, node, feature, value):
        """Insert a new node into the tree using recursion."""
        if value < node.value: # If the value is less than the current node
            if node.left is None: # If the left child is None
                node.left = IrisNode(feature, value) # Insert as the left child
            else: # If the left child is not None
                self._insert_recursive(node.left, feature, value) # Recur on the left child
        else: # If the value is greater than or equal to the current node
            if node.right is None: # If the right child is None
                node.right = IrisNode(feature, value) # Insert as the right child
            else: # If the right child is not None
                self._insert_recursive(node.right, feature, value) # Recur on the right child

    def in_order_traversal(self, node):
        """Perform in-order traversal of the tree."""
        if node: # If the node is not None
            self.in_order_traversal(node.left)  # Recur on the left child
            print(f"{node.feature}: {node.value}") 
            self.in_order_traversal(node.right) # Recur on the right child

iris_tree = IrisTree()
for _, row in iris_df.iterrows():
    iris_tree.insert('petal length (cm)', row['petal length (cm)'])

print("In-Order Traversal of Iris Tree:")
iris_tree.in_order_traversal(iris_tree.root)
# Classify flowers based on sepal and petal characteristics
```

### üöÄ **MapReduce**

MapReduce is a **distributed data processing framework** designed to handle massive datasets across multiple machines efficiently. It splits tasks into two main phases: **Map** and **Reduce**.

üõ†Ô∏è **MapReduce Workflow**

1. **Map Phase ($\textbf{map}$)**  
   - Processes each data record independently.  
   - Outputs intermediate **key-value pairs** $(k_1, v_1)$.

2. **Shuffle and Sort Phase ($\textbf{shuffle \& sort}$)**  
   - Groups intermediate key-value pairs by their **key** $(k_1)$.
   - Ensures that all values corresponding to the same key are processed together.

3. **Reduce Phase ($\textbf{reduce}$)**  
   - Processes grouped data from the **shuffle phase**.
   - Outputs the final result $(k_2, v_2)$.

##### **Mathematical Representation**

- **Map Function:**  
$$
map(k_1, v_1) \rightarrow [(k_2, v_2)]
$$

- **Reduce Function:**  
$$
reduce(k_2, [v_2]) \rightarrow (k_3, v_3)
$$

‚úÖ **Key Advantages of MapReduce**
- **Scalability:** Handles large datasets across distributed systems.
- **Fault Tolerance:** Recovers from node failures.
- **Parallelization:** Efficient use of multi-node architecture.
- **Abstraction:** Focus on *what* to compute, not *how* to parallelize.

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

iris_df['species'] = iris_df['species'].apply(lambda x: iris.target_names[x])

# Map Phase
def mapper(df):
    mapped_data = []
    for _, row in df.iterrows():
        mapped_data.append((row['species'], 1))
    return mapped_data

mapped_data = mapper(iris_df)
print("Mapped Data (First 5 Rows):")
print(mapped_data[:5])
```

```python
from collections import defaultdict

# Shuffle and Sort Phase
def shuffle_and_sort(mapped_data):
    grouped_data = defaultdict(list)
    for key, value in mapped_data:
        grouped_data[key].append(value)
    return grouped_data

grouped_data = shuffle_and_sort(mapped_data)
print("Grouped Data:")
print(grouped_data)
```

```python
# Reduce Phase
def reducer(grouped_data):
    reduced_data = {}
    for key, values in grouped_data.items():
        reduced_data[key] = sum(values)
    return reduced_data

reduced_data = reducer(grouped_data)
print("Reduced Data (Species Count):")
print(reduced_data)
```

```python
# All in one function
def mapreduce_species_count(df):
    mapped = mapper(df)
    grouped = shuffle_and_sort(mapped)
    reduced = reducer(grouped)
    return reduced

species_count = mapreduce_species_count(iris_df)
print("Final Species Count (MapReduce Result):")
for species, count in species_count.items():
    print(f"{species}: {count}")
```
