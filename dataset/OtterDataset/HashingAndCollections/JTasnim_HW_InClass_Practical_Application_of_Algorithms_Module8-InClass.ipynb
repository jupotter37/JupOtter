**Problem1-Leetcode#217-Contains Duplicate-Easy**

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

- Example 1:
    - Input: nums = [1,2,3,1]
    - Output: true

- Example 2:
    - Input: nums = [1,2,3,4]
    - Output: false
- Example 3:
    - Input: nums = [1,1,1,3,3,4,3,2,4,2]
    - Output: true

Problem1-Leetcode#217-Contains Duplicate
Explanation:

num_set is used because it stores only unique elements and provides O(1) average-time complexity for lookup and insertion

loop through each number n in the input list nums
for each number n, check if it already exists in num_set
if it does, it means the number has appeared before, so we return True immediately
if n is not already in the set, add it to num_set to keep track of the numbers encountered so far

if the loop completes without finding any duplicates, it means all numbers in the list are unique, so we return False


Time Complexity:  O(n)
Space Complexity: O(n)


**Problem2-Leetcode#1-Two Sum-Easy**

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

- Example 1:
    - Input: nums = [2,7,11,15], target = 9
    - Output: [0,1]
    - Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

- Example 2:
    - Input: nums = [3,2,4], target = 6
    - Output: [1,2]

- Example 3:
    - Input: nums = [3,3], target = 6
    - Output: [0,1]


Problem2-Leetcode#1-Two Sum-Easy
Explanation:

ind is a dictionary that maps numbers as keys to their indices as values in the list nums

in loop use enumerate(nums) to iterate over both the index i and the value n in the nums array
compute the complement of n as target - n
if the complement exists in ind, it means the current number n and its complement sum to the target,
then return the indices of these two numbers: the index of the complement (ind[target - n]) and the current index (i)

if the complement is not found, store the current number n in ind with its index i as the value

Time Complexity:  O(n)
Space Complexity: O(n)

**Problem3-Leetcode#146-LRU Cache-Medium**

Design a data structure that follows the constraints of a Least Recently Used (LRU) cache.

Implement the LRUCache class:

LRUCache(int capacity) Initialize the LRU cache with positive size capacity.\
int get(int key) Return the value of the key if the key exists, otherwise return -1.\
void put(int key, int value) Update the value of the key if the key exists. Otherwise,\
add the key-value pair to the cache. If the number of keys exceeds the capacity from \
this operation, evict the least recently used key.

The functions get and put must each run in O(1) average time complexity.

 

- Example 1:

    - Input ["LRUCache", "put", "put", "get", "put", "get", "put", "get", "get", "get"]\
    [[2], [1, 1], [2, 2], [1], [3, 3], [2], [4, 4], [1], [3], [4]]
    - Output
    [null, null, null, 1, null, -1, null, -1, 3, 4]

    - Explanation
        - LRUCache lRUCache = new LRUCache(2);
        - lRUCache.put(1, 1); // cache is {1=1}
        - lRUCache.put(2, 2); // cache is {1=1, 2=2}
        - lRUCache.get(1);    // return 1
        - lRUCache.put(3, 3); // LRU key was 2, evicts key 2, cache is {1=1, 3=3}
        - lRUCache.get(2);    // returns -1 (not found)
        - lRUCache.put(4, 4); // LRU key was 1, evicts key 1, cache is {4=4, 3=3}
        - lRUCache.get(1);    // return -1 (not found)
        - lRUCache.get(3);    // return 3
        - lRUCache.get(4);    // return 4


Problem3-Leetcode#146. LRU Cache-Medium
Explanation:
 
ListNode Class:
each ListNode represents a key-value pair in the cache
it has pointers to the next and previous nodes, enabling efficient addition and removal

LRUCache Initialization:

capacity: maximum number of key-value pairs the cache can hold
dic: a dictionary to map keys to their corresponding nodes in the linked list
head and tail: dummy nodes marking the start and end of the doubly linked list. This simplifies adding and removing nodes

get(key):

check existence, if key is not in dic, return -1

move to most recently used,
remove the node from its current position using remove
add it to the tail using add
return the value stored in the node

put(key, value):

update existing key,
if the key already exists, remove its current node to update it with the new value

add new node,
create a new ListNode with the key and value, and add it to the tail
store the node in the dictionary

evict least recently used key,
if the cache exceeds capacity, remove the node at the head (least recently used) and delete its entry from the dictionary

add(node):

adds a new node to the tail, marking it as the most recently used

remove(node):

removes the given node from its current position in the linked list by updating its neighbors

**Problem5-Leetcode#706- Design HashMap-Easy**

Design a HashMap without using any built-in hash table libraries.

Implement the MyHashMap class:

MyHashMap() initializes the object with an empty map.\
void put(int key, int value) inserts a (key, value) pair into the HashMap.\
If the key already exists in the map, update the corresponding value.
int get(int key) returns the value to which the specified key is mapped,\
or -1 if this map contains no mapping for the key.\
void remove(key) removes the key and its corresponding value\
if the map contains the mapping for the key.
 

- Example 1:

    - Input ["MyHashMap", "put", "put", "get", "get", "put", "get", "remove", "get"]\
          [[], [1, 1], [2, 2], [1], [3], [2, 1], [2], [2], [2]]
    - Output [null, null, null, 1, -1, null, 1, null, -1]

    - Explanation
        - MyHashMap myHashMap = new MyHashMap();
        - myHashMap.put(1, 1); // The map is now [[1,1]]
        - myHashMap.put(2, 2); // The map is now [[1,1], [2,2]]
        - myHashMap.get(1);    // return 1, The map is now [[1,1], [2,2]]
        - myHashMap.get(3);    // return -1 (i.e., not found), The map is now [[1,1], [2,2]]
        - myHashMap.put(2, 1); // The map is now [[1,1], [2,1]] (i.e., update the existing value)
        - myHashMap.get(2);    // return 1, The map is now [[1,1], [2,1]]
        - myHashMap.remove(2); // remove the mapping for 2, The map is now [[1,1]]
        - myHashMap.get(2);    // return -1 (i.e., not found), The map is now [[1,1]]


Problem5-Leetcode#706. Design HashMap-Easy
Explanation:

MyHashMap Class:

keyRange: the size of the hash_table is set to 769 (a prime number) to minimize collisions and evenly distribute keys
hash_table: an array where each element is an instance of the Bucket class
Each Bucket handles collisions for its index

Hash Function:

computes the index in the hash_table for a given key using the modulo operation (key % keyRange)

put(key, value):

Hash the Key: compute the index for the key using hash_func
insert or update: call the update method of the Bucket at the computed index to either insert the key-value pair or update its value if the key already exists

get(key):

Hash the Key: compute the index using hash_func.
retrieve value: call the get method of the Bucket at the computed index to return the value associated with the key, or -1 if the key does not exist

remove(key):

Hash the Key: compute the index using hash_func
delete jey: call the remove method of the Bucket at the computed index to remove the key-value pair

Bucket Class:

each Bucket contains a list bucket to store key-value pairs

update(key, value): 
update value: if the key exists in the bucket, update its value
insert Key: if the key does not exist, append the new key-value pair to the bucket

get(key):

traverses the bucket to find the key and returns its value if found. Otherwise, returns -1

remove(key):

traverses the bucket to find the key and deletes the corresponding key-value pair if found
