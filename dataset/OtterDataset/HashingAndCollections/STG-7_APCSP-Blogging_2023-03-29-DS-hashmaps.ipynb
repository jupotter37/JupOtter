## What is a Hashtable/Hashmap?

> A hashtable is a data structure that with a collection of key-value pairs, where each key maps to a value, and the keys must be unique and hashable.

- In Python there is a built in hashtable known as a **Dictionary.**

> The primary purpose of a hashtable is to provide efficient lookup, insertion, and deletion operations. When an element is to be inserted into the hashtable, a hash function is used to map the key to a specific index in the underlying array that is used to store the key-value pairs. The value is then stored at that index. When searching for a value, the hash function is used again to find the index where the value is stored.

> The key advantage of a hashtable over other data structures like arrays and linked lists is its average-case time complexity for lookup, insertion, and deletion operations.

- The typical time complexity of a hashtable is **O(1) â€” a constant time.** 


## What is Hashing and Collision?

> Hashing is the process of mapping a given key to a value in a hash table or hashmap, using a hash function. The hash function takes the key as input and produces a hash value or hash code, which is then used to determine the index in the underlying array where the value is stored. The purpose of hashing is to provide a quick and efficient way to access data, by eliminating the need to search through an entire data structure to find a value.

> However, it is possible for two different keys to map to the same hash value, resulting in a collision. When a collision occurs, there are different ways to resolve it, depending on the collision resolution strategy used.

> Python's dictionary implementation is optimized to handle collisions efficiently, and the performance of the dictionary is generally very good, even in the presence of collisions. However, if the number of collisions is very high, the performance of the dictionary can degrade, so it is important to choose a good hash function that minimizes collisions when designing a Python dictionary.

## What is a Set?

```python
# Creating a set using set() function
my_set = set([1, 2, 3, 2, 1])
print(my_set)  

# What do you notice in the output?
# The output elements are not ordered in the same way as they appeared in the original list. This is because sets are an unordered collection of elements.
# The original list [1, 2, 3, 2, 1] contains duplicates (i.e., 1 and 2 are repeated). However, when the list is converted to a set using set(), the duplicates are automatically removed and only the unique elements remain. Therefore, the output set contains only the unique elements of the original list, which are 1, 2, and 3.

# Why do you think Sets are in the same tech talk as Hashmaps/Hashtables?
# Sets and hash tables (also known as hash maps) are related data structures in computer science because they both rely on hash functions to store and retrieve data efficiently.
# Hash tables allow storage and retrieval of key-value pairs while sets are a collection of unique elements without values.
```

## Solution 1 - Making it a Set 

**In this solution, we use the set() function to convert the list of producers to a set, which automatically removes duplicates. Then we convert it back to a list before assigning it to the producer key in the lover_album dictionary.**

```python
lover_album["producer"] = list(set(['Taylor Swift', 'Jack Antonoff', 'Joel Little', 'Taylor Swift', 'Louis Bell', 'Frank Dukes']))
print(lover_album)
```

## Solution 2 - Checking for duplicates before adding to the list
**In this solution, we iterate through the list of producers and check if each producer is already in the unique_producers list. If it is not, we add it to the list. Finally, we assign the unique_producers list to the producer key in the lover_album dictionary.**

```python
producers = ['Taylor Swift', 'Jack Antonoff', 'Joel Little', 'Taylor Swift', 'Louis Bell', 'Frank Dukes']
unique_producers = []
for producer in producers:
    if producer not in unique_producers:
        unique_producers.append(producer)

lover_album["producer"] = unique_producers
print(lover_album)
```

## Solution 1 - Input validation
**Currently, the code does not check if the user input is valid or not. A user can enter anything, and the code will try to look up that key in the dictionary, which may not exist. To improve this, we can add input validation to ensure that the user inputs a valid key.**


```python
def search():
    while True:
        search = input("What would you like to know about the album?")
        if search.lower() not in lover_album:
            print("Invalid search term. Please try again.")
        else:
            print(lover_album[search.lower()])
            break

search()
```

## Solution 2 - Using .get() with a default value
**Currently, the code checks if a key exists in the dictionary using an if statement. An alternative approach is to use the .get() method with a default value. This method returns the value associated with the key if it exists, and a default value (in this case, "Invalid Search") if it doesn't.**

```python
def search():
    search = input("What would you like to know about the album?")
    print(lover_album.get(search.lower(), "Invalid Search"))

search()
```
