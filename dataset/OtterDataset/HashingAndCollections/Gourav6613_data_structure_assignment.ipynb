```python
# 1. Discuss string slicing and provide examples.

# ans:-  String slicing is a technique in Python used to access a range of characters within a string. 
#        This is achieved by specifying a start index, an end index, and an optional step value. The general syntax for string slicing is:
#        string[start:stop:step]
#        Here’s a breakdown of each part:

#        start: The index where the slice begins (inclusive).
#        stop: The index where the slice ends (exclusive).
#        step: The step size or stride between each index in the slice (optional, default is 1).
# Examples:-
# 1. Slicing with start and stop:
text = "Hello, World!"
print(text[0:5])  # Output: Hello
# This slice starts at index 0 and stops before index 5.

# 2. Slicing with start, stop, and step: 
print(text[0:5:2])  # Output: Hlo
# This slice starts at index 0, stops before index 5, and takes every 2nd character.

# 3. Slicing with negative indices:
print(text[-6:])  # Output: World!
#This slice starts 6 characters from the end and goes to the end of the string.

# 4. Slicing with omitted start or stop:
print(text[:5])  # Output: Hello
print(text[7:])  # Output: World!
print(text[:])   # Output: Hello, World!
#If start is omitted, it defaults to the beginning of the string. If stop is omitted, it defaults to the end of the string. 
#If both are omitted, it includes the entire string.

# 5. Slicing with negative step:
print(text[::-1])  # Output: !dlroW ,olleH
#This slice reverses the string by starting from the end and moving backwards.

```

```python
# 2. Explain the key features of lists in Python.

# ans:- Lists in Python are versatile and widely used data structures that allow you to store and manipulate collections of items.
#       Here are the key features of lists in Python:

# 1. Ordered Collection:
#    Lists maintain the order of elements as they are added. This means that elements can be accessed by their position (index) 
#    in the list.
numbers = [10, 20, 30, 40]
print(numbers[0])  # Output: 10

# 2. Mutable
#    Lists are mutable, which means that you can modify their content after they are created. You can change elements, add new elements,
#    or remove elements. 
numbers[1] = 25
print(numbers)  # Output: [10, 25, 30, 40]

numbers.append(50)
print(numbers)  # Output: [10, 25, 30, 40, 50]
fruits = ["apple", "banana", "cherry", "date"]
print(fruits[1])    # Output: banana
print(fruits[-1])   # Output: date
print(fruits[1:3])  # Output: ['banana', 'cherry']
print(fruits[:2])   # Output: ['apple', 'banana']

# 
numbers.remove(25)
print(numbers)  # Output: [10, 30, 40, 50]

# 3. Dynamic Size
#    Lists in Python can grow and shrink dynamically as elements are added or removed. 
#    There is no need to declare the size when the list is created it.
numbers = []
numbers.append(1)
numbers.append(2)
numbers.append(3)
print(numbers)  # Output: [1, 2, 3]

# 4. Heterogeneous Elements
#    Lists can contain elements of different data types, including other lists.
mixed = [1, "Hello", 3.14, [1, 2, 3]]
print(mixed)  # Output: [1, 'Hello', 3.14, [1, 2, 3]]

# 5. Indexing and Slicing
#    Lists support indexing and slicing to access subsets of elements.
fruits = ["apple", "banana", "cherry", "date"]
print(fruits[1])    # Output: banana
print(fruits[-1])   # Output: date
print(fruits[1:3])  # Output: ['banana', 'cherry']
print(fruits[:2])   # Output: ['apple', 'banana']

# 6. Comprehensions
#    List comprehensions provide a concise way to create lists. 
#    They consist of brackets containing an expression followed by a for clause.
squares = [x**2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]

# 7. Methods
#    Python lists come with a variety of built-in methods that facilitate common operations, 
#    such as appending, inserting, removing, and sorting elements.
numbers = [3, 1, 4, 1, 5, 9]
numbers.append(2)
print(numbers)  # Output: [3, 1, 4, 1, 5, 9, 2]

numbers.sort()
print(numbers)  # Output: [1, 1, 2, 3, 4, 5, 9]

numbers.reverse()
print(numbers)  # Output: [9, 5, 4, 3, 2, 1, 1]

# 8. Iterability
#    Lists are iterable, meaning you can loop through the elements using loops.
for fruit in ["apple", "banana", "cherry"]:
    print(fruit)
# Output:
# apple
# banana
# cherry

# 9. Membership Testing
#    it can be checked for the presence of an element in a list using the in keyword
numbers = [1, 2, 3, 4, 5]
print(3 in numbers)  # Output: True
print(6 in numbers)  # Output: False

#    Summary
#    Lists in Python are powerful and flexible. Their key features include order preservation, mutability, dynamic sizing, 
#    support for heterogeneous elements, indexing and slicing capabilities, comprehensions for concise construction, a variety of useful
#    methods, iterability, support for nested lists, and easy membership testing. These features make lists a fundamental and indispensable 
#    part of Python programming.

```

```python
# 3. Describe how to access, modify, and delete elements in a list with examples. 

# ans:- Accessing Elements in a List:
# 1. By Index:
#    You can access elements in a list by their index. Indexing starts at 0 for the first element.
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # Output: apple
print(fruits[1])  # Output: banana
print(fruits[-1])  # Output: cherry (last element)

# 2. By Slicing:
#    Slicing allows you to access a range of elements.
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])  # Output: [20, 30, 40]
print(numbers[:3])  # Output: [10, 20, 30]
print(numbers[2:])  # Output: [30, 40, 50]
print(numbers[::2])  # Output: [10, 30, 50] (every second element)

#       Modifying elements in a list:
# 1. Changing an Element by Index:
#    You can modify an element by assigning a new value to its index.
fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

# 2. Modifying Multiple Elements using Slicing:
#    You can use slicing to modify a range of elements.
numbers = [10, 20, 30, 40, 50]
numbers[1:4] = [21, 31, 41]
print(numbers)  # Output: [10, 21, 31, 41, 50]

# 3. Using List Methods:
#    append(): Adds an element to the end of the list.
#    insert(): Adds an element at a specified index.
#    extend(): Adds multiple elements to the end of the list.
fruits.append("date")
print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'date']

fruits.insert(1, "banana")
print(fruits)  # Output: ['apple', 'banana', 'blueberry', 'cherry', 'date']

fruits.extend(["elderberry", "fig"])
print(fruits)  # Output: ['apple', 'banana', 'blueberry', 'cherry', 'date', 'elderberry', 'fig']

#       Deleting Elements in a List:
# 1. Using del:
#    You can delete an element or a slice of elements using the del statement.
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
del fruits[1]
print(fruits)  # Output: ['apple', 'cherry', 'date', 'elderberry']

del fruits[1:3]
print(fruits)  # Output: ['apple', 'elderberry']

# 2. Using remove():
#    The remove() method deletes the first occurrence of a value
fruits = ["apple", "banana", "cherry", "date", "banana"]
fruits.remove("banana")
print(fruits)  # Output: ['apple', 'cherry', 'date', 'banana']

# 3. Using pop():
#    The pop() method removes an element at a specified index and returns it. If no index is specified, 
#    it removes and returns the last element.
fruits = ["apple", "banana", "cherry", "date"]
popped = fruits.pop(2)
print(popped)  # Output: cherry
print(fruits)  # Output: ['apple', 'banana', 'date']

last_fruit = fruits.pop()
print(last_fruit)  # Output: date
print(fruits)  # Output: ['apple', 'banana']

# 4. Using clear():
#    The clear() method removes all elements from the list.
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(fruits)  # Output: []       

#       Summary
#       Accessing: Use indexing and slicing to access elements.
#       Modifying: Change elements directly by index, use slicing for multiple elements, or use list methods like append(),
#       insert(), and extend().
#       Deleting: Use del to remove elements by index, remove() to delete by value, pop() to remove and return an element by index,
#       and clear() to remove all elements.
```

```python
# 4. Compare and contrast tuples and lists with examples.

# ans:- Tuples and lists are both sequence data types in Python that can store a collection of items.
#       However, they have distinct characteristics and serve different purposes. Here’s a comparison of tuples and lists, 
#       highlighting their similarities and differences:

#    1. Mutability
# a) Lists: Mutable, meaning their contents can be changed after creation.
# b) Tuples: Immutable, meaning their contents cannot be changed after creation.
#Eg.
# List (mutable)
fruits = ["apple", "banana", "cherry"]
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

# Tuple (immutable)
colors = ("red", "green", "blue")
# colors[1] = "yellow"  # This would raise a TypeError

# 2. Syntax
# a) Lists: Defined using square brackets [].
# b) Tuples: Defined using parentheses ().
#Eg.
# List
numbers = [1, 2, 3, 4]

# Tuple
coordinates = (10.0, 20.0)

# 3. Performance
# a) Lists: Generally slower for operations like iteration and accessing elements because of their dynamic nature.
# b) Tuples: Generally faster due to their immutability and fixed size, which allows for optimization.
#Eg.
# Tuples can be more efficient for fixed data sets that don't need modification

# 4. Use Cases
# a) Lists: Suitable for collections of items that may need to be modified, such as adding, removing, or changing elements.
# b) Tuples: Suitable for fixed collections of items, such as coordinates, or when data integrity is important (e.g., using tuples as 
#    dictionary keys).
#Eg.
# List use case
shopping_list = ["eggs", "milk", "bread"]
shopping_list.append("butter")

# Tuple use case
point = (4, 5)  # Representing a coordinate

# 5. Methods
# a) Lists: Have a variety of built-in methods like append(), remove(), sort(), etc.
# b) Tuples: Have fewer methods since they cannot be modified. Mainly support count() and index().
#Eg.
# List methods
numbers = [1, 3, 2, 4]
numbers.sort()
print(numbers)  # Output: [1, 2, 3, 4]

# Tuple methods
weekdays = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday")
print(weekdays.index("Wednesday"))  # Output: 2

# 6. Size and Memory
# a) Lists: Tend to consume more memory because they require extra space for new elements, allowing for dynamic resizing.
# b) Tuples: Use less memory, as they are fixed-size and immutable.
#Eg.
# Tuples are preferred when memory efficiency is crucial.

```

```python
# 5. Describe the key features of sets and provide examples of their use.

# ans:- Sets in Python are a collection data type that is unordered, mutable, and unindexed. They are commonly used for membership testing,
#       eliminating duplicate entries, and set operations like union and intersection. 
#       Here are the key features of sets and examples of their use:
#    Key Features of Sets:
# 1. Unordered Collection:
#  Sets do not maintain any particular order for the elements. The elements have no index, and their order may change over time.
fruits = {"apple", "banana", "cherry"}
print(fruits)  # Output could be in any order, like {'banana', 'apple', 'cherry'}

# 2. Mutable:
# Sets can be modified after creation. You can add and remove elements,
# but the elements themselves must be immutable (e.g., numbers, strings, tuples).
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits)  # Output: {'apple', 'banana', 'cherry'}

# 3. No Duplicate Elements
# Sets automatically handle duplicate elements. If you try to add an existing element, it will not appear twice.
numbers = {1, 2, 3, 4, 4, 4}
print(numbers)  # Output: {1, 2, 3, 4}

# 4. Unindexed
# Sets do not support indexing or slicing since they are unordered. You cannot access elements by position.
fruits = {"apple", "banana", "cherry"}
fruits[0]  # This would raise a TypeError

# 5. Efficient Membership Testing
# Sets support mathematical set operations like union, intersection, difference, and symmetric difference.
fruits = {"apple", "banana", "cherry"}
print("apple" in fruits)  # Output: True
print("orange" in fruits)  # Output: False

# Sets are an excellent choice when you need to store a collection of unique elements, perform set operations, 
# or check for membership efficiently. They provide a powerful tool for various applications in data analysis, computation,
# and logic-based tasks
```

```python
# 6. Discuss the use cases of tuples and sets in Python programming. 

# ans:- Use Cases of Tuples in Python:
# 1. Fixed Collections of Data
# Tuples are ideal for representing a collection of data that should not change. Examples include geographical coordinates, 
# RGB color values, or configuration settings.
#Eg.
coordinates = (40.7128, -74.0060)  # (latitude, longitude)
rgb_color = (255, 0, 0)  # Red color in RGB

# 2. Return Multiple Values from a Function
# Tuples can be used to return multiple values from a function in a single call
#Eg.
def get_min_max(numbers):
    return (min(numbers), max(numbers))

result = get_min_max([1, 2, 3, 4, 5])
print(result)  # Output: (1, 5)

# 3. Dictionary Keys
# Tuples can be used as keys in dictionaries because they are immutable. 
# This is useful when you need a compound key, such as a pair of coordinates or a name and date.
#Eg.
locations = {
    (40.7128, -74.0060): "New York",
    (34.0522, -118.2437): "Los Angeles"
}
print(locations[(40.7128, -74.0060)])  # Output: New York

# 4. Data Integrity
# When data should remain constant throughout the program, tuples provide a way to ensure immutability, 
# preventing accidental modifications.
config = ("localhost", 8080, "root", "password")

# 5. Packing and Unpacking
# Tuples are often used for packing and unpacking multiple values. This feature is handy for simultaneous assignment and swapping variables.
#Eg.
x, y, z = (1, 2, 3)
print(x, y, z)  # Output: 1 2 3

a, b = 5, 10
a, b = b, a
print(a, b)  # Output: 10 5

#       Use Cases of Sets in Python:
# 1. Removing Duplicates:
# Sets automatically remove duplicate elements. They are useful for extracting unique items from a list or other iterable.
#Eg.
items = [1, 2, 2, 3, 4, 4, 5]
unique_items = set(items)
print(unique_items)  # Output: {1, 2, 3, 4, 5}

# 2. Membership Testing
# Sets provide fast membership testing due to their underlying hash table structure. This is useful for scenarios where you need to frequently
# check if an item exists in a collection.
#Eg.
allowed_users = {"Alice", "Bob", "Charlie"}
user = "David"
if user in allowed_users:
    print("Access granted")
else:
    print("Access denied")
    
# 3. Mathematical Set Operations:
# Sets support operations like union, intersection, difference, and symmetric difference. These operations are useful in data analysis, 
# such as finding common elements or unique elements between datasets.
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Union
print(set_a | set_b)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection
print(set_a & set_b)  # Output: {3, 4}

# Difference
print(set_a - set_b)  # Output: {1, 2}

# Symmetric Difference
print(set_a ^ set_b)  # Output: {1, 2, 5, 6}

# 4. Unordered Data Storage:
# Sets are useful when the order of elements is not important. For instance, they can be used to store tags, categories, or labels.
tags = {"python", "programming", "development"}

# 5. Efficient Data Management
# Sets can efficiently manage and manipulate large collections of data, especially when it comes to tasks like eliminating duplicates,
# merging collections, and finding differences.
students_set1 = {"Alice", "Bob", "Charlie"}
students_set2 = {"Bob", "David", "Emma"}

all_students = students_set1 | students_set2
print(all_students)  # Output: {'Alice', 'Bob', 'Charlie', 'David', 'Emma'}

# Summary
# Tuples are used when you need a fixed collection of items, where the order and immutability of the items are important.
# They are ideal for use as dictionary keys, function returns, and for ensuring data integrity.

# Sets are best for scenarios where the uniqueness of elements is crucial, where you need efficient membership testing, or when performing
# mathematical set operations. They are suitable for tasks like removing duplicates, managing unordered data, and performing set algebra.
```

```python
# 7. Describe how to add, modify, and delete items in a dictionary with examples.

# ans:- Dictionaries in Python are unordered collections of items. Each item is stored as a key-value pair, where the key is unique 
# and immutable, and the value can be any data type. Here's how to add, modify, and delete items in a dictionary:

# 1. Adding Items to a Dictionary
# To add a new key-value pair to a dictionary, you can use the assignment operator (=) with the new key and value.
# Initializing an empty dictionary
person = {}

# Adding key-value pairs
person["name"] = "John Doe"
person["age"] = 30
person["city"] = "New York"

print(person)  # Output: {'name': 'John Doe', 'age': 30, 'city': 'New York'}
# can also use the update() method to add multiple key-value pairs at once
person = {"name": "John Doe"}
person.update({"age": 30, "city": "New York"})

print(person)  # Output: {'name': 'John Doe', 'age': 30, 'city': 'New York'}

# 2. Modifying Items in a Dictionary
# To modify an existing value, you can assign a new value to an existing key using the assignment operator (=). 
# If the key exists, its value will be updated
person = {"name": "John Doe", "age": 30, "city": "New York"}

# Modifying the value of an existing key
person["age"] = 31
person["city"] = "Los Angeles"

print(person)  # Output: {'name': 'John Doe', 'age': 31, 'city': 'Los Angeles'}

# 3. Deleting Items from a Dictionary
# There are several ways to delete items from a dictionary:
# Using del: The del statement removes a key-value pair by key. If the key does not exist, it raises a KeyError.
person = {"name": "John Doe", "age": 32, "city": "San Francisco"}
del person["age"]

print(person)  # Output: {'name': 'John Doe', 'city': 'San Francisco'}

# Using pop(): The pop() method removes the key and returns its value. It also raises a KeyError if the key does not exist. 
# You can specify a default value to return if the key is not found.
age = person.pop("age", None)  # None is returned if "age" does not exist
print(age)      # Output: 32
print(person)   # Output: {'name': 'John Doe', 'city': 'San Francisco'}

# Using clear(): The clear() method removes all items from the dictionary.
person = {"name": "John Doe", "city": "San Francisco"}
person.clear()

print(person)  # Output: {}

# Summary
# Adding Items: Use the assignment operator or the update() method.
# Modifying Items: Reassign values using the assignment operator or the update() method.
# Deleting Items: Use del, pop(), popitem(), or clear() methods.
```

```python
# 8. Discuss the importance of dictionary keys being immutable and provide examples.

# ans:- In Python, dictionary keys must be immutable because they are used as identifiers to store and retrieve data efficiently. 
#       Immutability ensures that the hash value of a key remains constant, which is crucial for the performance and reliability 
#       of hash tables, the underlying data structure of dictionaries. Here’s a detailed discussion on the importance of immutability 
#       for dictionary keys, along with examples:
# Importance of Immutability for Dictionary Keys:
# 1. Consistency in Hashing
# Dictionaries use a hash function to map keys to specific locations in the underlying hash table. If a key's value changes,
# its hash value would also change, making it impossible to find the stored value. Immutability ensures that the key's 
# hash value remains constant throughout its lifetime in the dictionary.
#Eg.
d = {}
key = [1, 2, 3]  # A list, which is mutable
# d[key] = "value"  # This would raise a TypeError because lists are not hashable

# 2. Uniqueness of Keys
# Immutability helps maintain the uniqueness of keys. If keys were mutable, they could be altered after being added to the dictionary, 
# potentially leading to key collisions and data inconsistency.

# 3. Performance
# The efficiency of lookups, insertions, and deletions in a dictionary relies on the hash function. If keys were mutable and could change
# after insertion, it would slow down operations and require rehashing.

# 4. Data Integrity
# Using immutable keys prevents accidental changes, ensuring that the data associated with those keys remains accessible and accurate.

#       Examples of Immutable and Mutable Types:
#    Immutable Types (Allowed as Keys):
# Strings: Immutable and commonly used as keys.
# Numbers (int, float): Immutable and can be used as keys.
# Tuples: Immutable as long as they contain only immutable elements.
# Frozenset: An immutable version of sets, suitable for keys
#Eg.
d = {
    "name": "Alice",
    42: "The Answer",
    (1, 2): "coordinates",
    frozenset([1, 2, 3]): "immutable set"
}
print(d["name"])  # Output: Alice
print(d[42])      # Output: The Answer
print(d[(1, 2)])  # Output: coordinates

#    Mutable Types (Not Allowed as Keys):
# Lists: Mutable, as their contents can be changed.
# Dictionaries: Mutable, and therefore cannot be used as keys.
# Sets: Mutable and cannot be used as keys
#Eg.
d = {[1, 2]: "value"}  # Raises TypeError: unhashable type: 'list'
d = {{1: "one"}: "value"}  # Raises TypeError: unhashable type: 'dict'

# Summary
# Immutability of dictionary keys is crucial for maintaining data integrity, ensuring efficient operations, and preventing accidental
# modifications. Immutable types such as strings, numbers, tuples, and frozensets are allowed as dictionary keys, while mutable types like 
# lists, dictionaries, and sets are not. This immutability requirement ensures that the keys remain consistent, unique, and reliably 
# hashable throughout their use in a dictionary.
```
