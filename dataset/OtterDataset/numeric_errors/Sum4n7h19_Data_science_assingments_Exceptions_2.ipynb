```python
#Q2
def python_exception(exception_class,indent=0):
    print(""* indent + exception_class.__name__)
    for sub_class in exception_class.__subclasses__():
        python_exception(sub_class,indent+1)
print("python_exception hirerchay:")
python_exception(BaseException)
```

```python
#Q3
"""there are mainly three errors are defined in Arithmetic error,they are:"""
#floating point error
#zerodivision error
try:
    
    divisor_error =1/0
    print(divisor_error)
except ArithmeticError as i:
    print(f"{i},{i.__class__}")
    
```

```python
#overflow error
m=3.0

try:
    for i in range(1,100):
        m=m**i
except ArithmeticError as e:
    print(f"{e},{e.__class__}")
    
```
