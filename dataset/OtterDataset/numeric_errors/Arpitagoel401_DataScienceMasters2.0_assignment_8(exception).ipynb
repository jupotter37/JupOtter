Q2. Write a python program to print Python Exception Hierarchy.

```python
def exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        exception_hierarchy(subclass, indent + 2)

exception_hierarchy(Exception)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# LookupError class in Python is the base class for exceptions that occur when a key or index is not found. 
# It is a subclass of the Exception class.

my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d'] 
    
except KeyError as e:
    print(e)

    
my_list = [1, 2, 3]    
try:
    value = my_list[4]
except IndexError as e:
    print(f"Error: {e}")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
# ImportError
try:
    import non
except ImportError as e:
    print(f"Import Error: {e}")

#ModuleNotFoundError
try:
    from non import module_x
except ModuleNotFoundError as e:
    print(f"Module not found Error: {e}")
```
