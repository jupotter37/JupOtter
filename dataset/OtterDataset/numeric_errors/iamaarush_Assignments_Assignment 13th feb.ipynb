```python
# Example of overflow error
""" Overflow error occurs when the numeric calculation is too big"""
import logging
logging.basicConfig(filename = "assignment.log" , level = logging.ERROR)

n= 4.3
try:
    
    for i in range(1, 1000000):
        n= n**i
        
except OverflowError as e:
    
    logging.error("this will show you the overflow error{}".format(e))
    
    print("this will show you the overflow error"  ,e)
```

```python
# Question 4

"""The lookuperror exception forms a base class for all exceptions that are raised when an index or a key is not found
for a sequence or dictionary respectively . we can use lookup exception class to handle both index and key error
exception classes."""

# Example of index error
"""This error occur when an attempt is made to access an item in a list at an index which is out of range."""
try:
    
    l=[1,2,3,4]
    print(l[5])
except IndexError as e:   # we can use lookuperror also in the place of indexerror.

    print("this will show an index error  " ,(e))
```

```python
# example of key error
"""This error occurs when an attempt is made to access an item in dictionary that does not exist."""
try:
    d={"key" : "value" , 45:"fish" , "name" : "aarush"}
    print(d["key2"])

except LookupError as e:    # just to show that we can also use this base class for exception. 
    print("we can use lookup error also to handle key or index error" , e)
    
except KeyError as e:
    print("This will show us a key error as the key we want to access does not exist  " , e)
```

```python
# Question 6 

# Best practices for exception handling in python are -

1 . Try to use specific exception always and not the super class Exception.

2. print or log always a proper message so that whoever is seeing your code can properly understand it.

3. always try to log your errors in the place of print so that in the future whoever want to see the code can see it 
and understands it and it will help in debugging.

4. Do not write multiple exception only use the proper exception whichever the code needs at that time.

5 . Always do documentation of the code with proper meaning so that in future developers understand that code
and makes it easier for them to maintain and update it.
```
