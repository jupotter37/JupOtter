4. Why LookupError class is used? Explain with an example KeyError and IndexError.

In Python, the LookupError class is the base class for all exceptions that occur when a key or index is not found in a sequence or mapping.

```python
#keyerror
d = {'a': 1, 'b': 2}

try:
    value = d['c']
except KeyError as e:
    print("Error:", e)
```

```python
#indexerror
l = [1, 2, 3]

try:
    value = l[3]
except IndexError as e:
    print("Error:", e)
```

5. Explain ImportError. What is ModuleNotFoundError?

- In Python, ImportError is an exception that is raised when an imported module, package, or object cannot be found or imported. This can occur when the module name is misspelled, the module is not installed, or the module is not accessible from the current working directory or Python path.

- In Python 3.6 and later versions, a new exception ModuleNotFoundError was introduced to specifically handle cases when an imported module cannot be found. This exception is a subclass of ImportError and provides a more informative error message.
