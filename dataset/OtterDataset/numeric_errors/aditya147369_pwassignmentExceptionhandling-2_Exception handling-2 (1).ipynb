Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans - The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

The Python KeyError is an exception that occurs when an attempt is made to access an item in a dictionary that does not exist. The key used to access the item is not found in the dictionary, which leads to the KeyError

```python
pylenin_info = {'name': 'Lenin Mishra',
                'age': 28,
                'language': 'Python'}
user_input = input('What do you want to learn about Pylenin==> ')

try:
    print(f'{user_input} is {pylenin_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

An IndexError means that your code is trying to access an index that is invalid. This is usually because the index goes out of bounds by being too large. For example, if you have a list with three items and you try to access the fourth item, you will get an IndexError.

```python
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans - An ImportError in Python signals a failure during the import process, typically when the interpreter cannot locate or load a module. This could happen due to various reasons, such as a typo in the module name, an incorrect path, or a missing dependency. It's a broad exception that encompasses different import-related issues.

ModuleNotFoundError is a more specific subclass of ImportError introduced in Python 3.6. It's raised exclusively when the named module cannot be found. This distinction helps pinpoint the exact cause of the import failure, making debugging more straightforward. In older Python versions, ModuleNotFoundError scenarios would also fall under the general ImportError umbrella.
