os ramos except são pesquisados na mesma ordem em que aparecem no código;

não se deve usar mais de um ramo except com um determinado nome de exceção;

o número de diferentes ramos except é arbitrário – a única condição é que, se você usar try,
deve colocar pelo menos um except (nomeado ou não) depois dele;

a palavra-chave except não deve ser usada sem um try precedente;

se qualquer um dos ramos except for executado, nenhum outro ramo será visitado;

se nenhum dos ramos except especificados corresponder à exceção levantada, a exceção permanece não tratada

se existir um ramo except não nomeado (sem um nome de exceção), ele deve ser especificado por último.

## Lembre-se:

- a ordem dos ramos importa!

- não coloque exceções mais gerais antes das mais específicas;

- isso tornará as exceções específicas inacessíveis e inúteis;

- além disso, isso fará com que seu código fique desorganizado e inconsistente;

- o Python não gerará nenhuma mensagem de erro sobre essa questão.

### assert expression3

Como funciona?

Avalia a expressão;
Se a expressão for avaliada como True, ou um valor numérico diferente de zero, ou uma string não vazia, ou qualquer outro valor diferente de None, não fará mais nada;
Caso contrário, levantará automática e imediatamente uma exceção chamada AssertionError (nesse caso, dizemos que a asserção falhou).


Como pode ser usada?

Você pode querer colocá-la no seu código onde deseja estar absolutamente seguro contra dados evidentemente errados, e onde não tem certeza absoluta de que os dados foram cuidadosamente examinados antes (por exemplo, dentro de uma função usada por outra pessoa);
Levantar uma exceção AssertionError protege seu código de produzir resultados inválidos e mostra claramente a natureza da falha;
As asserções não substituem as exceções nem validam os dados – elas são seus complementos.

```python
import math

x = float(input("Enter a number: "))
assert x >= 0.0 # se ocorrer o erro esse trecho será exibido

x = math.sqrt(x)

print(x)

# A instrução do Python raise ExceptionName pode levantar uma exceção sob demanda.
# A mesma instrução, mas sem ExceptionName, pode ser usada apenas dentro do ramo except,
# e levanta a mesma exceção que está sendo tratada no momento.

# A instrução assert expression do Python avalia a expressão e levanta a exceção AssertError
# quando a expressão é igual a zero, uma string vazia ou None. 
# Você pode usá-la para proteger algumas partes críticas do seu código contra dados devastadores.
```

```python
# **ArithmeticError**
# Localização: BaseException ← Exception ← ArithmeticError
# Descrição: uma exceção abstrata que inclui todas as exceções causadas
# por operações aritméticas como divisão por zero ou domínio inválido de um argumento.

# **AssertionError**
# Localização: BaseException ← Exception ← AssertionError
# Descrição: uma exceção concreta levantada pela instrução assert 
# quando seu argumento é avaliado como False, None, 0 ou uma string vazia.


from math import tan, radians
angle = int(input('Enter integral angle in degrees: '))

# We must be sure that angle != 90 + k * 180
assert angle % 180 != 90
print(tan(radians(angle)))
```

```python
# **BaseException**
# Localização: BaseException
# Descrição: a mais geral (abstrata) de todas as exceções do Python
# – todas as outras exceções estão incluídas nesta; pode-se dizer 
# que os seguintes dois ramos except são equivalentes: except: e except BaseException

# **IndexError**
# Localização: BaseException ← Exception ← LookupError ← IndexError
# Descrição: uma exceção concreta levantada quando você tenta acessar 
# um elemento inexistente de uma sequência (por exemplo, um elemento de uma lista).


# The code shows an extravagant way
# of leaving the loop.
the_list = [1, 2, 3, 4, 5]
ix = 0
do_it = True
 
while do_it:
    try:
        print(the_list[ix])
        ix += 1
    except IndexError:
        do_it = False
 
print('Done')
```

```python
# **KeyboardInterrupt**
# Localização: BaseException ← KeyboardInterrupt
# Descrição: uma exceção concreta levantada quando o usuário
# usa um atalho de teclado projetado para terminar a execução de um programa 
# (Ctrl-C na maioria dos sistemas operacionais); se o tratamento dessa exceção 
# não levar à terminação do programa, ele continuará sua execução.

# This code cannot be terminated
# by pressing Ctrl-C.
 
from time import sleep
 
seconds = 0
 
while True:
    try:
        print(seconds)
        seconds += 1
        sleep(1)
    except KeyboardInterrupt:
        print("Don't do that!")
```

```python
# OverflowError
# Localização: BaseException ← Exception ← ArithmeticError ← OverflowError
# Descrição: uma exceção concreta levantada quando uma operação produz
# um número grande demais para ser armazenado com sucesso.

# The code prints subsequent
# values of exp(k), k = 1, 2, 4, 8, 16, ...

from math import exp

ex = 1

try:
    while True:
        print(exp(ex))
        ex *= 2
except OverflowError:
    print('The number is too big.')
```

## LAB Leitura segura de inteiros

Escrever uma função capaz de receber valores inteiros e 
verificar se eles estão dentro de um intervalo especificado.

A função deve:

Aceitar três argumentos: um prompt, um limite mínimo aceitável e um limite máximo aceitável;
    
Se o usuário inserir uma string que não seja um valor inteiro, a função deve exibir a mensagem
"Erro: entrada incorreta" e pedir ao usuário para inserir o valor novamente;

Se o usuário inserir um número que esteja fora do intervalo especificado,
a função deve exibir a mensagem "Erro: o valor não está dentro do intervalo permitido (mín..máx)" 
e pedir ao usuário para inserir o valor novamente;

Se o valor de entrada for válido, retorná-lo como resultado.

É assim que a função deve reagir à entrada do usuário:

Digite um número de -10 a 10: 100
Erro: o valor não está dentro do intervalo permitido (-10..10)
Digite um número de -10 a 10: asd
Erro: entrada incorreta
Digite um número de -10 a 10: 1
O número é: 1

```python
def read_int(prompt, min, max):
    while True:
        try:
            prompt = int(input(('Digite um número de -10 a 10: ')))
            if  -10 <= prompt <= 10:
                return prompt
            else:
                print('Erro: o valor não está dentro do intervalo permitido (-10..10)')
        except ValueError:
            print('Erro: entrada incorreta')                  

v = read_int("Enter a number from -10 to 10: ", -10, 10)
print("O número é:", v)
    
```

```python
def read_int(prompt, min, max):
    ok = False
    while not ok:
        try:
            value = int(input(prompt))
            ok = True
        except ValueError:
            print("Error: wrong input")
        if ok:
            ok = value >= min and value <= max
        if not ok:
            print("Error: the value is not within permitted range (" + str(min) + ".." + str(max) + ")")
    return value;


v = read_int("Enter a number from -10 to 10: ", -10, 10)

print("The number is:", v)
```
