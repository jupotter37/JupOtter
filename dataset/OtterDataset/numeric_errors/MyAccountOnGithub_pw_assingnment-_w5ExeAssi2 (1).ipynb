```python
# Q2. Write a python program to print Python Exception Hierarchy.
"""Printing Python's built-in exception hierarchy programmatically is a bit tricky because Python doesn't 
provide a built-in way to get all exceptions. However, we can generate a reasonable approximation by 
exploring the subclasses of the built-in BaseException class, which is the base class for all built-in 
exceptions"""
def print_exception_hierarchy(exception_type, indent=0):
    print(' '*indent + exception_type.__name__)
    for subclass in exception_type.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(BaseException)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
"""The LookupError base class is used for exceptions that are raised when a key or index used on a mapping 
or sequence is invalid: i.e., when it does not exist. The two most common examples of this are KeyError and 
IndexError.
keyerror: Is occur when the key is print is not present in dictionary then the errror is occur 
Indexerror:when in list we print out of list range """
try:
    d={'a':1,'b':2,'c':3}
    print(d['d'])
except KeyError as e:
    print("the keyerror occur ",e)
```
