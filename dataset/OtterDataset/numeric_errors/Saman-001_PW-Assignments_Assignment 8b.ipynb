##### Question 3

The ArithmeticError class in Python is the base class for all errors associated with arithmetic operations. It has three subclasses:

1. OverflowError-Raised when an operation results in a value that is too large to be represented in the machine's memory.
2. ZeroDivisionError-Raised when an attempt is made to divide by zero.
3. FloatingPointError-Raised when an operation results in a floating-point error.

```python
i=1
try:
    f = 3.0**i
    for i in range(100):
        print (i, f)
        f = f ** 2
except OverflowError as err:
    print ('Overflowed after ', f, err)
```

```python
#zero division error 

dividend = 10
divisor = 0

try:
    result = dividend / divisor
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

##### Question 4


The LookupError class is used to represent errors that occur when a key or index is not found in a sequence or mapping. The LookupError class has two subclasses:

1. KeyError: Raised when a key is not found in a dictionary.
2. IndexError: Raised when an index is out of range for a sequence.

```python
# Key Error

my_dict = {"a": 1, "b": 2}

try:
    value = my_dict["c"]
except KeyError as e:
    print("Error",e)
```

# Index Error

1. Use try/except blocks to handle specific exceptions. Don't catch all exceptions, as this can mask other errors. Instead, catch specific exceptions that you expect your code to raise, and handle them appropriately.
2. Log exceptions. This will help you to track down and fix errors when they occur.
3. Display helpful error messages to users. Don't just print the exception message to the console. Instead, try to explain what went wrong and what the user can do to fix it.
4. Don't ignore exceptions. If an exception is raised, don't just ignore it. This can lead to unexpected program behavior or even crashes.
5. Reraise exceptions when necessary. If you can't handle an exception yourself, you can reraise it so that it can be handled by higher-level code.
