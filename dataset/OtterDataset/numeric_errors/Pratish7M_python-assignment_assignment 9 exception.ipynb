```python
'''     Q5. Explain ImportError. What is ModuleNotFoundError?       '''


#Answer :



'''
ImportError is an exception that is raised when an import statement fails to import a module or when 
there is an error in the process of importing a module. It is a broad exception that encompasses various 
import-related errors.

ModuleNotFoundError is a subclass of ImportError that specifically indicates that a module could not be 
found or imported. It was introduced in Python 3.6 to provide a more specific exception for cases where 
a module is not found.'''


try :
    import non_existance 
except ModuleNotFoundError as e :
    print(f"Module not found: {e.name}")

try :
    import non_existance 
except ImportError as e :
    print("error found , not found any", e.name)
```

```python
'''   Q6. List down some best practices for exception handling in python.    '''



#Answer :


'''Be specific in catching exceptions: Catch only the exceptions that you expect and know how to handle. 
Avoid catching broad exceptions like Exception or BaseException unless you have a good reason to do so. 
Catching specific exceptions allows for more precise handling and avoids obscuring unexpected errors.

Use multiple except blocks: When you need to handle different exceptions differently, use separate except 
blocks for each exception. This allows you to provide specific error handling logic based on the type of exception.

Use a finally block: If you have cleanup or resource release operations that need to be performed regardless 
of whether an exception occurs or not, use a finally block. The code within the finally block is executed no
 matter what, ensuring proper cleanup of resources.

Avoid bare except blocks: Avoid using bare except blocks without specifying the exception type. It can make 
debugging and troubleshooting difficult since any exception raised within the try block will be caught, 
including unexpected ones. Instead, catch specific exceptions or at least catch Exception if you want to 
handle all exceptions.

Handle exceptions at the right level: Handle exceptions at the appropriate level in your code. Determine whether 
it's best to handle an exception locally within a function or propagate it up to the calling code. Catching exceptions 
too early or too late can affect the clarity and effectiveness of error handling.

Provide meaningful error messages: Include informative error messages when raising exceptions or handling them. 
Clear and descriptive error messages can help with debugging and troubleshooting. Include relevant information 
that helps identify the cause of the exception.'''
```
