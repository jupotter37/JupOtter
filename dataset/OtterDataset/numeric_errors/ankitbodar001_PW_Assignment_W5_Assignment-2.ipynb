```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python
The ArithmeticError Exception is the base class for all errors associated with arithmetic operation.
ArithmeticError
--> ZeroDivisionError
--> OverflowError
--> FloatingPointError
```

```python
#Example 1 - Handling ZeroDivisionError
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
'''As you can see, the ArithmeticError exception class is able to handle ZeroDivisionError 
exception. The e.__class__ method tells you that it was a ZeroDivisionError.'''
```

```python
#Example 2 - Handling OverflowError
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
-->This is the base class for those exceptions that are raised when a key or index used on a 
mapping or sequence is invalid or not found.
```

```python
#An IndexError is raised when a sequence reference is out of range
try:
    foo = ['a', 's', 'd', 'f', 'g']
    print (foo[5])
except IndexError as e:
    print (e)
```

```python
##An KeyError is raised when Key in not found
pylenin_info = {'name': 'Lenin Mishra',
                'age': 28,
                'language': 'Python'}
user_input = input('What do you want to learn about Pylenin==> ')

try:
    print(f'{user_input} is {pylenin_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

```python
Q6. List down some best practices for exception handling in python.
```

```python
1. Use Exceptions for Exceptional Cases
2. Don’t Swallow the Exception
3. Catch Specific Exceptions
4. Always Clean Up Resources in a Finally Block
5. Avoid Raising Generic Exceptions
6. Raise Custom Exceptions
7. Define Your Own Exception Hierarchy
8. Document All Exceptions Thrown by a Function
9. Provide Contextual Information When Raising an Exception
10. Write Tests to Ensure That Exceptions Are Raised Correctly
```
