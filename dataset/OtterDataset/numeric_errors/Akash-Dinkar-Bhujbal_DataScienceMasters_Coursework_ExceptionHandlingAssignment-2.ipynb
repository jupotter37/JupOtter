# Q1. Explain why we have to use the Exception class while creating a custom Exception.

Ans.:

-   In Python, all exceptions are derived from the base class `BaseException`.
-   For custom exceptions, it is recommended to inherit from the `Exception` class. This provides consistency in exception handling and allows you to catch your custom exceptions along with other built-in exceptions.
-   Using the `Exception` class also ensures that your custom exception behaves like a standard exception, making it easier for developers to understand and handle.

# Q2. Write python program to print Python Exception Hierarchy.

```python
# Python program to print Python Exception Hierarchy

def print_exception_hierarchy(exception, level=0):
    print("  " * level + f"{exception.__name__}")
    for subclass in exception.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

# Print the exception hierarchy
print_exception_hierarchy(BaseException)
```

# Q3. What errors are defined in the ArithmeticError class? Explain any two with an Example.

Ans.:
The ArithmeticError class in Python represents errors that occur during arithmetic operations. Two common errors are:

1.    `ZeroDivisionError`: Raised when division or modulo by zero is encountered.
2.    `FloatingPointError`: This error occurs when a floating-point operation results in an undefined or unrepresentable value.

```python
# ZeroDivisionError Example
try:
    result = 10 / 0  # Division by zero
except ZeroDivisionError as e:
    print(f"Error: {e}")

# FloatingPointError Example
try:
    result = 1.0 / 0.0  # Attempting to divide by zero in floating-point arithmetic
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

# Q4. Why `LookupError` class is used? Explain with an example `KeyError` and `IndexError`.

Ans.:

-    The `LookupError` class is the base class for exceptions that occur when a key or index is not found. It includes `IndexError` and `KeyError`.
-    `IndexError` is raised when an index is out of range, and `KeyError` is raised when a dictionary key is not found.

```python
# IndexError Example
my_list = [1, 2, 3]
try:
    value = my_list[5]  # Index out of range
except IndexError as e:
    print(f"Error: {e}")

# KeyError Example
my_dict = {"name": "John", "age": 25}
try:
    value = my_dict["gender"]  # Key not found
except KeyError as e:
    print(f"Error: {e}")
```

# Q5. Explain `ImportError`. What is `ModuleNotFoundError`?

Ans.:

- `ImportError` is raised when an import statement fails to find the module or attribute being imported.
- `ModuleNotFoundError` is a subclass of `ImportError` and is specifically raised when the specified module is not found.

```python
# ImportError Example
try:
    import non_existent_module  # Trying to import a non-existent module
except ImportError as e:
    print(f"Error: {e}")

# ModuleNotFoundError Example
try:
    from non_existent_module import non_existent_function  # Trying to import a non-existent function
except ModuleNotFoundError as e:
    print(f"Error: {e}")
```

# Q6. List down some best preactices for exception handling in python.
Ans.:

**Best Practices:**
1. Use Always a Specific Exception
2. Print Always a Proper Message
3. Always Try to Log Your Errors
4. Always Avoid Writing Multiple Exception Handling
5. Document All the Error / Cleanup All the Resources
6. Ensure proper cleanup even if an exception occurs before the with block

```python
# Use Always a Specific Exception
try:
    10/0
except Exception as e:
    print(e)

try:
    10/0
except ZeroDivisionError as e:
    print(e)

# Print Always a Proper Message
try:
    10/0
except ZeroDivisionError as e:
    print("I am trying to handle a zerodivision error: ", e)

# Always Try to Log Your Errors
import logging
logging.basicConfig(filename="error.log", level=logging.ERROR)

try:
    10/0
except ZeroDivisionError as e:
    logging.error("I am trying to handle a zerodivision error: {}".format(e))

# Always Avoid Writing Multiple Exception Handling
try:
    10/0
except ZeroDivisionError as e:
    logging.error("I am trying to handle a zerodivision error: {}".format(e))

# Document All the Error / Cleanup All the Resources
try:
    with open("test101.txt", 'w') as f:
        f.write("This is my data to file")
except FileNotFoundError as e:
    logging.error("File not found: {}".format(e))
except Exception as e:
    logging.error("An unexpected error occurred: {}".format(e))
finally:
    # Ensure proper cleanup even if an exception occurs before the with block
    if 'f' in locals():
        f.close()
```
