```python
#Q4
'''The LookupError class is a built-in exception class in Python that is used as the base class for various other exception classes related to lookup operations. 
The main purpose of LookupError is to handle situations where an index or key is not found in a sequence or dictionary.

Two common subclasses of LookupError are KeyError and IndexError. Here's an example of each:'''

#KeyError:
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}
try:
    value = my_dict['grape']
except KeyError:
    print("KeyError: 'grape' is not in my_dict")
'''In this example, we are trying to access the value of the key 'grape' in a dictionary. However, the key does not exist in the dictionary, so a KeyError is raised.
We catch the exception using a try-except block and print a custom error message.'''

#Indexerror
my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError:
    print("IndexError: my_list index out of range")

'''In this example, we are trying to access the value of the element at index 3 in a list. However, the list only has elements at indices 0, 1, and 2, so an IndexError is raised.
We catch the exception using a try-except block and print a custom error message.'''
```

```python
#Q5
'''In Python, the ImportError exception is raised when a module, package, or dependency cannot be found or imported properly.
This can occur for various reasons such as an incorrect module name or a missing dependency.'''
try:
    import Numpy
except ImportError:
    print("Numpy module not found")
'''In this example, we are trying to import the numpy module. If the module is not found or cannot be imported for any reason, an ImportError is raised.
We catch the exception using a try-except block and print a custom error message.'''
```
