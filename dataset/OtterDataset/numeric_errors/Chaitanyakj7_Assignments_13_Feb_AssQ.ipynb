Answer 3) OverflowError, ZeroDivisionError(DivisionByZero, DivisionUndefined), FloatingPointError are the
errors defined in the ArithmeticError class

```python
try:
    print(10/0)
except ZeroDivisionError as e:
    print("Error handled is:-", e)

import math

try:
    print(math.exp(1000))
except OverflowError as e:
    print("Error handled:-",e)
```

Answer 4) LookupError(StandardError) is used as a base class for 
exceptions raised when a sequence or dictionary type doesn't 
contain a given index or key.

```python
try:
    d = {"one":1, "two":2}
    d["three"]
except KeyError as e:
    print(e)

try:
    l = [1, 2, 3]
    l[100]
except IndexError as e:
    print(e)
```

Answer 5) ImportError occurs when the Python program tries to import
module which does not exist.

```python
try:
    import hello
except ImportError as e:
    print(e)

try:
    import hello
except ModuleNotFoundError as e:
    print(e)
```
