```python
#Q4 Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
# The LookupError class in Python is a base class for exceptions raised when a key or index lookup fails.
# KeyError is Raised when a dictionary key is not found.
# IndexError is Raised when an index is out of range for a sequence (like in list)
```

```python
try:
    d={"key":"sudh",1:[1,2,3]}
    print(d["key2"])
except KeyError as e:
    print(e)
```

```python
try:
    l=[1,2,3]
    print(l[6])
except IndexError as e:
    print(e)
    
```

```python
# Q5Explain ImportError. What is ModuleNotFoundError?
```

```python
# ImportError is raised in Python when an import statement fails to load a module or a part of a module.
# ModuleNotFoundError is a subclass of ImportError that specifically indicates that a module could not be found during an import attempt
```

```python
try:
    import sudh
except ImportError as e:
    print(e)
```

```python
try:
    import sudh
except ModuleNotFoundError as e:
    print(e)
```
