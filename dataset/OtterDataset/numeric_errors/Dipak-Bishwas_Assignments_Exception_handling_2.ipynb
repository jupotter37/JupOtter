```python
# Q2. Write a python program to print Python Exception Hierarchy.
import logging

logging.basicConfig(filename='list.log', level=logging.INFO, format='%(message)s')

def log_exception_hierarchy(base_exception, indent=0):

    logging.info('  ' * indent + base_exception.__name__)
    
    for subclass in base_exception.__subclasses__():
        log_exception_hierarchy(subclass, indent + 1)

logging.info("Python Exception Hierarchy:")
log_exception_hierarchy(Exception)


```

```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
# Ans: ZeroDivisionError: Occurs when dividing by zero.
# OverflowError: Occurs when the result of an arithmetic operation exceeds the limits of the data type.
```

```python
logging.basicConfig(filename='error.log', level=logging.ERROR, format='%(asctime)s - %(levelname)s - %(message)s')

try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error("An error occurred: %s", e)
```

```python
import sys

try:
    result = sys.maxsize + 1
except OverflowError as e:
    logging.error("An error occurred: %s", e)
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

# Ans: 
# The LookupError class is a general error type for problems related to finding things in lists or dictionaries. If you try to get an item from a list or dictionary with a wrong key or index, Python raises specific errors like KeyError or IndexError, both of which are types of LookupError. This means you can catch all these specific lookup errors at once using LookupError in your error handling code.

```

```python

my_dict = {"name": "Alice", "age": 25}

try:
    value = my_dict["address"] 
except KeyError as e:
    logging.error("KeyError: %s", e)
```

```python
my_list = [1, 2, 3]

try:
    value = my_list[5] 
except IndexError as e:
    logging.error("IndexError: %s", e)
```
