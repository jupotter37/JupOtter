### 1 . Why Exception class is used as base class to create Custom Exceptions ?

In python, it is important to inherit from Exception class when creating a custom exceprion.because of following reasons :

- It is easier to understand for developers to see what kind of exceptions are used

- inheriting from existing Exceptions will allow you to take advantage of error handling mechanism

- you can also customize your custom Exception behaviour . you can add or remove custom attributes to customize how we handle the exception

- it is helpful to document the code and help developers understand that code is intended to represent Exception class

### 3 . Errors in ArthematicError class and two examples

Arthematicerror is a general exception that occurs when there is an error in calculation of numbers.
some of the errors available in this Exception class are as follows :

1. ZeroDivisionError
2. OverflowError
3. FloatingPointError
4. TimeoutError
5. UnderflowError

```python
## demonstrating example of ZeroDivisionError
try:
    a=10
    b=0
    c=a/b
except ZeroDivisionError as e:
    print(e)
    
```

```python
## demonstrating example of OverflowError
try:
    import math
    print(math.exp(1000))
    
except OverflowError as e:
    print(e)
```

### 4 . LookupError and example of keyerror and index error

LookupError is form of base class for exceptions that occur when a lookup or indexing operation fails.
this can handle both "KeyError" and "indexError" when used along with "try" block and "except" clause.
     

Following are the example of 

1. KeyError
2. IndexError

KeyError occurs when we are trying to call a key that is not available or not defined inside a dictionary . following is the example of KeyError

```python
## Example of KeyError
try:
    d = {'name':'Madhusudhan','age':24,'Role':'Data scientist','Institute':'pwskills'}
    d['mentor']
except KeyError as e :
    print("Key error, cannot find key: ",e)
```

Index error occurs when we are calling for an index position that is exceeding the actual available data items inside the list. following is example of index error

```python
## example of index error
try:
    lst = [1,2,3,4,5]
    lst[13]
except IndexError as e:
    print(e)
```
