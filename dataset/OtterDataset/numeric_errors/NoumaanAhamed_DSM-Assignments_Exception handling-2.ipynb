```python
import sys 

def print_exception_class(exception_class,indent = 0):
    print(" "*indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_class(subclass,indent + 4)

print_exception_class(Exception)
```

Question 4

```python
#The LookupError class in Python is a base class for exceptions related to lookup operations, such as indexing or key-based lookup
```

```python
my_dict = {"name": "John", "age": 25}

try:
    value = my_dict["occupation"]
except KeyError as error:
    print("Error:", str(error))
```

```python
my_list = [1, 2, 3]

try:
    value = my_list[5]
except IndexError as error:
    print("Error:", str(error))
```
