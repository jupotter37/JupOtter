## Built-in modules

```python
import math
x = 1.55
y = -1.55

# get the largest integer less than x
print(math.floor(x)) # 1
print(math.floor(y)) # -2
# get the smallest integer greater than x
print(math.ceil(x)) # 2
print(math.ceil(y)) # -1
# drop fractional part of x
print(math.trunc(x)) # 1, equivalent to math.floor for positive numbers
print(math.trunc(y)) # -1, equivalent to math.ceil for negative numbers
```

```python
# Localization
import locale

print('Format numbers numbers for given locales.\n')

sample_locales = [('INDIA', 'as_IN'), ('USA', 'en_US'), ('France', 'fr_FR'), ('Spain', 'es_ES'), ('Portugal', 'pt_PT'), ('Poland', 'pl_PL')]

print('{:>10} {:>10} {:>15} {:>20}  {:>20}'.format('Locale', 'Integer', 'Float', 'Positive balance', 'Negative balance'))
for name, loc in sample_locales:
    locale.setlocale(locale.LC_ALL, loc)

    print('{:>10}'.format(name), end=' ')
    print(locale.format_string('%10d', 123456, grouping=True), end=' ')
    print(locale.format_string('%15.2f', 123456.78, grouping=True), end=' ')
    print('{:>20}  {:>20}'.format(locale.currency(1234.56), locale.currency(-1234.56)))
```

```python
# Requests is very helpful to make HTTP calls
import requests
# Pprint module lets us pretty-print arbitrary data structures in Python 
# to make them prettier, well-formatted, and more readable.
from pprint import pprint
  
def fetchUserDetails(userId): 
    url = "https://jsonplaceholder.typicode.com/users"
    resp = requests.get(url, params = {'id': userId}) 
    return resp.json() 
  
# calling geocode function 
data = fetchUserDetails(1) 
  
# printing json response
print('Printing without foramating json response\n')
print(data)
print('\n\nPrinting with foramating json response\n')
pprint(data) 
```
