Q5. Explain ImportError. What is ModuleNotFoundError?
Ans- Import error is the python message that is raised when import statement faces trouble on succesfully importing the specific module . This is due to the invalid syntax.
ModuleNotFoundError is raised when Python cannot find the module specified in the import statement. It can also occur when the module you're trying to import whicch does not exist or cannot be located in the sys.path or any other paths specified for module search.
ModuleNotFoundError is a more specific error that deals exclusively with the issue of a module not being found, whereas ImportError is more general and can be raised for other import-related issues.

```python
try:
    import name      ##his is raising import error as if there is problem in import statement
except ImportError as e:
    print(e)
```

```python
try:
    import name      ##his is raising ModuleNotFoundError as there is no module named name
except ModuleNotFoundError as e:
    print(e)
```
