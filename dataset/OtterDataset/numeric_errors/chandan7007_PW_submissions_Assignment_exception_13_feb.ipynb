OverflowError
ZeroDivisionError

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

exception LookupError
This is the base class for those exceptions that are raised when a key or index used on a mapping or 

sequence is invalid or not found. The exceptions raised are :
a)KeyError
b)IndexError

```python
try: 
    a = [1, 2, 3] 
    print (a[3]) 
except LookupError: 
    print ("Index out of bound error.")
else: 
    print ("Success")
```

```python
try: 
    a = {"a":1,"b":2}
    print (a["c"]) 
except KeyError: 
    print ("key error.")
else: 
    print ("Success")
```

exception ImportError

An ImportError is raised when the import statement is unable to load a module or when the “from list” in
