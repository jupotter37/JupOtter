```python
#q1
Inheritance from a Base Class:

The Exception class serves as the base class for all built-in exceptions in Python. When you create a custom exception class, you're essentially defining a new type of exception that inherits the behavior and attributes of the base Exception class.
Standardized Exception Handling:

By inheriting from Exception, you ensure that your custom exception follows the standard conventions and behaviors expected of exceptions in Python. This includes having attributes like args to store error messages and methods like __str__ for string representation.
Interoperability:

Inheriting from Exception allows your custom exception to be used interchangeably with built-in exceptions in exception handling code. This means you can catch your custom exception in the same way you catch any other exception in Python.
```

```python
#q2
def print_exception_hierarchy(exception_class, level=0):
    print('  ' * level + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

# Start with the base Exception class
print_exception_hierarchy(Exception)
```

```python
#q3
#ArithmeticError is a base class for arithmetic-related exceptions in Python. It itself is not raised in specific situations, but rather serves as a superclass for more specific arithmetic exceptions.

#Two commonly used exceptions that are subclasses of ArithmeticError are:

#ZeroDivisionError: This exception is raised when you try to divide a number by zero.
try:
    result = 10 / 0  # Attempting to divide by zero
except ZeroDivisionError as e:
    print(f"Error: {e}")
    
   # OverflowError: This exception is raised when an arithmetic operation exceeds the maximum limit of a numeric type.

import sys
try:
    result = sys.maxsize + 1  # Trying to exceed the maximum value of the integer type
except OverflowError as e:
    print(f"Error: {e}")

```

```python
#q4
#LookupError is a base class for exceptions that occur when a key or index used to access a value in a collection (like a dictionary or a list) is not found. It serves as the parent class for more specific lookup-related exceptions in Python.

#Two commonly used exceptions that are subclasses of LookupError are KeyError and IndexError.

#KeyError:

#This exception is raised when you try to access a dictionary with a key that doesn't exist.
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']  # Attempting to access a non-existent key
except KeyError as e:
    print(f"Error: {e}")
#IndexError:

#This exception is raised when you try to access an index that is out of range in a sequence (like a list or a tuple).
my_list = [1, 2, 3]
try:
    value = my_list[3]  # Attempting to access an index beyond the list length
except IndexError as e:
    print(f"Error: {e}")
```

```python
#q5
#ImportError is an exception in Python that is raised when the import statement fails to find a module, or when there's an issue with importing a module. This can happen for various reasons, such as a typo in the module name, the module not being installed, or a circular import.

#ModuleNotFoundError is a subclass of ImportError introduced in Python 3.6. It specifically indicates that a module could not be found or imported. It provides a more descriptive error message than a generic ImportError, making it easier to identify the module that could not be imported.
try:
    import non_existing_module  # Attempting to import a non-existent module
except ModuleNotFoundError as e:
    print(f"Error: {e}")
```
