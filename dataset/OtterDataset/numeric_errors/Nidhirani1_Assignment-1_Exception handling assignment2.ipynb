```python
Q6. List down some best practices for exception handling in python.
Ans:=Exception handling is crucial for writing robust and maintainable Python code. Here are some best practices to follow:

1. Use Specific Exceptions
Be Specific: Catch specific exceptions rather than a generic Exception. This makes your error handling more precise and helps in diagnosing issues more effectively.


try:
    result = 1 / 0
except ZeroDivisionError:
    print("Cannot divide by zero.")
Avoid Bare Except: Avoid using except: without specifying the exception type, as it catches all exceptions including those you may not expect or handle correctly.


try:
    # Some code
except Exception as e:
    # Handle exception
2. Handle Exceptions at the Right Level
Granular Handling: Handle exceptions at the level where you can actually do something meaningful about them. For example, if you’re working with file operations, handle IOError close to the file handling code.


try:
    with open('file.txt') as f:
        data = f.read()
except FileNotFoundError:
    print("File not found.")
3. Use finally for Cleanup
Resource Management: Use the finally block to ensure that resources such as files or network connections are properly closed or cleaned up, regardless of whether an exception occurs.


try:
    f = open('file.txt')
    # Process file
except IOError:
    print("An error occurred.")
finally:
    f.close()
Context Managers: Prefer using context managers (with statements) for resource management, as they automatically handle cleanup.


with open('file.txt') as f:
    data = f.read()
4. Log Exceptions
Logging: Instead of just printing error messages, use Python’s logging module to record exceptions. This helps in tracking and debugging issues later.


import logging

logging.basicConfig(level=logging.ERROR)

try:
    result = 1 / 0
except ZeroDivisionError as e:
    logging.error("An error occurred: %s", e)
5. Reraise Exceptions When Necessary
Preserve Traceback: If you need to handle an exception but still want to propagate it up the stack, use raise without arguments inside an except block.


try:
    # Some code
except ValueError as e:
    # Handle it, but rethrow it to be handled further up
    raise
6. Avoid Empty except Clauses
Avoid Silence: An empty except block (i.e., except:) should be avoided because it swallows exceptions without any handling or notification, which can make debugging difficult.


try:
    # Some code
except SomeException:
    pass  # This should be avoided
7. Be Careful with Exception Handling in Loops
Handle Carefully: When handling exceptions inside loops, be cautious not to mask errors or inadvertently skip important iterations.


for item in items:
    try:
        process(item)
    except SomeException:
        handle_exception()
8. Document Your Exception Handling
Document Decisions: Clearly document why certain exceptions are being caught and how they are being handled. This improves code readability and maintainability.


try:
    # Code that might raise an exception
except SomeException:
    # Handle the exception in this way because...
9. Test Your Exception Handling
Unit Testing: Ensure that your exception handling code is covered by unit tests. This verifies that your application behaves as expected in the presence of errors.


import unittest

class TestMyFunction(unittest.TestCase):
    def test_exception(self):
        with self.assertRaises(ValueError):
            my_function()
```
