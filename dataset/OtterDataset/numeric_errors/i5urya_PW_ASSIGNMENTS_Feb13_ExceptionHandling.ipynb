Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
Ans :The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

A Python IndexError exception is what is raised when you try to access a index that is not present.

A Python KeyError exception is what is raised when you try to access a key that isn’t in a dictionary (dict).

You can use LookupError exception class to handle both IndexError and KeyError exception classes.

```python
import sys
try:
    foo = ['a', 's', 'd', 'f', 'g']
    print(foo[5])
except IndexError as e:
    print(f"{e}, {e.__class__}")
```

```python
try:
    d={1:[1,2,3,4,5],"key":"deepa"}
    d["key10"]
except KeyError as e:
    print(f"{e}, {e.__class__}")
```

Q6. List down some best practices for exception handling in python.

```python
#1)Always use a specific exception
try:
    a=20
    a/0
except ZeroDivisionError as e:   
    print(e)
```

```python
#2)Print always a proper message
try:
    a=20
    a/0
except ZeroDivisionError as e:   
    print("I have handle the zero division error")
```

```python
#3)Always try to log
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    a=20
    a/0
except ZeroDivisionError as e:
    logging.error("This is my Zero division exception which I'm handling {}".format(e))
```

```python
#4) Always avoid to write unnecessary exception
try:
    a=20
    a/0   
except FileNotFoundError as e:
    logging.error("This is file not found Exception  {}".format(e))
    
except AttributeError as e:
    logging.error("This is Attribute Error Exception  {}".format(e))
    
except ZeroDivisionError as e:
    logging.error("This is my Zero division exception which I'm handling {}".format(e))
```
