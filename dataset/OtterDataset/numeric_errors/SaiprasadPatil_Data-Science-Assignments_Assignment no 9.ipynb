Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

In Python, the ArithmeticError class is the base class for exceptions 
that are raised for arithmetic errors. It itself does not define anyspecific errors but serves as a superclass
for a group of related exception classes that deal with arithmetic operations.
Some of the commonly used exceptions derived from ArithmeticError include are,
ZeroDivisionError 
,FloatingPointError 
,OverflowError 
,UnderflowError 
,ArithmeticError 

```python
try :
    10/0
except ArithmeticError as a:
    print(a)
```

```python
import math as m
try:
    print(m.exp(1000))
except OverflowError as e:
    print(e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is used as a base class for exceptions that are raised when a lookup or indexing operation fails.
It serves as a superclass for a group of related exception classes that deal with lookup-related errors. 
The main purpose of the LookupError class is to provide a consistent interface and behavior for handling lookup failures.
it is used for Handling specific lookup errors

```python
try:
    dic={"name":"sai","email":"sai@gmail.com","ph_no":9348215}
    print(dic["name1"])
except KeyError as e :
    print(e)
```

```python
try:
    l=[1,2,3,4,5]
    l[6]
except IndexError as e:
    print(e)
```
