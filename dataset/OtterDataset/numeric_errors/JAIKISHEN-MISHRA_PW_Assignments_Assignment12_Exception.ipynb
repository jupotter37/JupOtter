```python
#Q2. Write a python program to print Python Exception Hierarchy.
#Ans:-
import inspect as ipt   
def tree_class(cls, ind = 0):    
    print ('-' * ind, cls.__name__)    
    for K in cls.__subclasses__():  
        tree_class(K, ind + 3)  
    
print ("The Hierarchy for inbuilt exceptions is: ")    
ipt.getclasstree(ipt.getmro(Exception))
tree_class(Exception)  
```

```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
#Ans:-
#1.ZeroDivisionError: This exception is raised when attempting to divide a number by zero.
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Division by zero")
```

```python
#2.OverflowError: This exception is raised when the result of an arithmetic operation is too large to be expressed within the range of the
#available numeric type.
import sys

try:
    result = sys.maxsize + 1
except OverflowError:
    print("Error: Result exceeds the maximum value")
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
#Ans:-The LookupError class in Python is a base class for exceptions that occur when a lookup or indexing operation fails.
#1.KeyError
my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']
except KeyError:
    print("Error: Key not found in dictionary")
#2.IndexError
my_list = [1, 2, 3]

try:
    value = my_list[3]
except IndexError:
    print("Error: Index out of range")
```

```python
#Q5. Explain ImportError. What is ModuleNotFoundError?
#Ans:-ImportError is an exception that is raised when an import statement fails to import a module.
#ModuleNotFoundError is a specific subclass of ImportError that is raised when a module cannot be found during the import process.Example:-
try:
    import jai
except ModuleNotFoundError:
    print("Error: Module not found")
```
