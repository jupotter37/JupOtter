Python's exception hierarchy is organized in a tree-like structure, with BaseException at the root. The Exception class serves as a base for other common exceptions like ValueError, TypeError, RuntimeError, etc. By inheriting from Exception, our custom exception becomes part of this hierarchy, which can be useful when we want to catch exceptions at different levels of specificity.

This is the reason why we inherit from the `Exception` class while creating our Custom Exception.

```python
#Q2.

def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    if issubclass(exception_class, BaseException):
        for subclass in exception_class.__subclasses__():
            print_exception_hierarchy(subclass, indent + 4)


# Start printing the exception hierarchy from the base class `BaseException`
print_exception_hierarchy(BaseException)
```

'KeyError` is raised when we try to access a dictionary using a key which doesnot exist in the dictionary. Here's an example:

```python
my_dict = {'a':1, "b":2, "c":3}
print(my_dict["D"])
```

Q5.

`ImportError` is a base exception class that is raised when an import statement fails for any reason. It can be raised when there is a problem with finding, loading, or executing a module or package. It is a general error that encompasses various import-related issues.


`ModuleNotFoundError` is a specific exception class that is derived from `ImportError`. It is raised when an import statement fails because the specified module or package cannot be found. This typically occurs when the module or package does not exist in the Python installation or is not accessible from the current working directory or the Python path.
