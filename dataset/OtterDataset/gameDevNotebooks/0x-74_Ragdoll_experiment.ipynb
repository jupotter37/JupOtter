```python
llm = Groq(model="llama3-groq-70b-8192-tool-use-preview", api_key=f"{os.getenv('groq')}")
```

```python
class Resource(BaseModel):
    """Data model for organizing technical resources"""
    title: str
    description: str
    is_video_or_blog_or_course: str
    programming_languages : str
    framework_used : str
    tags : str
```

```python
class ResourceValidator(BaseModel):
    """data model for validating extracted resources """
    has_page_loading_error:bool
```

```python
class Retagger(BaseModel):
    """data model for retagging and removing redundant tags and returning all tags in lowercase"""
    lowercase_cleaned_tags:list
```

```python
import csv

def write_to_csv(index, url,reason, file):
    with open(file, mode='a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([index, url, reason])  # Write index and URL as a comma-separated row
```

```python
def validate_resource(df):
    write_to_csv('index','url','reason', 'df_fails.csv')
    structured_llm = llm.as_structured_llm(output_cls=ResourceValidator)
    for index, row in df.iterrows():
        time.sleep(2)
        input_msg = ChatMessage.from_str(f"check if there was an error opening the following site from the given description {row['description']}")
        output = structured_llm.chat([input_msg]).raw.dict()
        if output['has_page_loading_error']:
            write_to_csv(index, row['url'],row['description'], 'df_fails.csv',)
                
```

```python
def retag(df):
    df1 = df.copy()
    structured_llm = llm.as_structured_llm(output_cls=Retagger)
    input_msg = ChatMessage.from_str(f"return cleaned tags from all redundant tags and repeated tags for example AI and artificial intelligence become one tag , also return all tags in lowercase here is the list of tags : \n{' '.join(df.tags.str.split(',').explode().unique().tolist())}")
    output = structured_llm.chat([input_msg]).raw.dict()
    cleaned_tags = output['lowercase_cleaned_tags']
    for index, row in df1.iterrows():
        time.sleep(5)
        input_msg = ChatMessage.from_str(f"""return a list of tags from {cleaned_tags} which is lowercase by analysing the following tags and removing reduntant tags in the following :\n"{row['tags']}""")
        output = structured_llm.chat([input_msg]).raw.dict()
        df1.at[index, 'tags'] = output['lowercase_cleaned_tags']
    return df1
```

```python
def write_to_file(url,file):
    f = open(file, "a")
    f.write(url)
    f.write('\n')
    f.close()
```

```python
def extract_technical_tags(url):
    try:
        print(f'processing :{url}')
        time.sleep(2)
        # Load web page content and convert HTML to text
        webpage_data = SimpleWebPageReader(html_to_text=True).load_data([url])

        # Use a structured LLM to parse the webpage content into a Resource object
        structured_llm = llm.as_structured_llm(output_cls=Resource)
        input_msg = ChatMessage.from_str(f"Generate technical tags for the following content {webpage_data[0]}")
        
        # Define a chat prompt to ask for technical tags
        output = structured_llm.chat([input_msg]).raw.dict()

        # Parse the response into JSON format and return it
        output["url"] = url
        return output

    except Exception as e:
        write_to_file(url)
        print(f"An error occurred: {e}")
        pass
```

```python
df = pd.DataFrame(columns=['title', 
                           'description',
                           'is_video_or_blog_or_course', 
                           'programming_languages', 
                           'framework_used', 
                           'tags', 
                           'url'])
```

```python
for url in urls:
    if re.search(r"arxiv", url):
        write_to_file(url)
        continue
    if re.search(r'youtu\.*be',url):
        write_to_file(url)
        continue
    ex = extract_technical_tags(url)
    df.loc[len(df)] = ex
```
