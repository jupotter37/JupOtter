```python
# Retrieve destrieux parcellation in fsaverage5 space from nilearn
from nilearn import datasets
from nilearn import plotting
from nilearn import surface
import nibabel as nib

parcellations = datasets.fetch_atlas_basc_multiscale_2015(version='asym')

print(parcellations['scale122'])

fsaverage = datasets.fetch_surf_fsaverage()

# The fsaverage dataset contains file names pointing to the file locations
print('Fsaverage5 pial surface of left hemisphere is at: %s' %
      fsaverage['pial_left'])
print('Fsaverage5 inflated surface of left hemisphere is at: %s' %
      fsaverage['infl_left'])
print('Fsaverage5 sulcal depth map of left hemisphere is at: %s' %
      fsaverage['sulc_left'])
```

```python
label = 65

networks_122 = nib.load(parcellations['scale122'])
networks_122.get_data()[networks_122.get_data()!=label] = 0
networks_122.get_data()[networks_122.get_data()==label] = 100

cmap = 'blue_red'

texture = surface.vol_to_surf(networks_122, fsaverage.pial_left, kind='ball', radius=20)
plotting.plot_surf_stat_map(fsaverage.infl_left, texture, hemi='left',
                            title='Surface left hemisphere lateral', colorbar=True, view='lateral',
                            threshold=1, vmax=200, bg_map=fsaverage.sulc_left, cmap=cmap)
plotting.show()

texture = surface.vol_to_surf(networks_122, fsaverage.pial_left, kind='ball', radius=20)
plotting.plot_surf_stat_map(fsaverage.pial_left, texture, hemi='left',
                            title='Surface left hemisphere lateral', colorbar=True, view='lateral',
                            threshold=1, vmax=200, bg_map=fsaverage.sulc_left, cmap=cmap)
plotting.show()

texture = surface.vol_to_surf(networks_122, fsaverage.pial_right, kind='ball', radius=20)
plotting.plot_surf_stat_map(fsaverage.infl_right, texture, hemi='right',
                            title='Surface left hemisphere lateral', colorbar=True, view='lateral',
                            threshold=1, vmax=200, bg_map=fsaverage.sulc_right, cmap=cmap)
plotting.show()

texture = surface.vol_to_surf(networks_122, fsaverage.pial_right, kind='ball', radius=20)
plotting.plot_surf_stat_map(fsaverage.pial_right, texture, hemi='right',
                            title='Surface left hemisphere lateral', colorbar=True, view='lateral',
                            threshold=1, vmax=200, bg_map=fsaverage.sulc_right, cmap=cmap)
plotting.show()
```

```python
plotting.plot_surf_roi(fsaverage['infl_left'], roi_map=parcellations['scale122'],
                       hemi='left', view='lateral',
                       bg_map=fsaverage['sulc_left'], bg_on_data=True,
                       darkness=.5)

plotting.show()
```

```python
texture = surface.vol_to_surf(networks_122, fsaverage.pial_right, kind='ball', radius=5)
plotting.plot_surf_stat_map(fsaverage.infl_right, texture, hemi='right',
                            title='Surface left hemisphere lateral', colorbar=True, view='ventral',
                            threshold=1, vmax=200, bg_map=fsaverage.sulc_left, cmap=cmap)
plotting.show()

texture = surface.vol_to_surf(networks_122, surface.load_surf_data('/home/tspisak/src/RPN-signature/data/atlas/cerebellum.gii'), kind='ball', radius=5)
plotting.plot_surf_stat_map(fsaverage.sulc_left, texture, hemi='right',
                            title='Surface left hemisphere lateral', colorbar=True, view='ventral',
                            threshold=1, vmax=200, bg_map=fsaverage.sulc_left, cmap=cmap)
plotting.show()
```
