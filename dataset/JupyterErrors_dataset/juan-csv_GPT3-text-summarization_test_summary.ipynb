```python
# summarization
def summary_gpt3(input_text):
    response = openai.Completion.create(
        engine="davinci-instruct-beta-v3",
        prompt=f"Given a transcript, get summary:\n\nTranscript: {input_text}\n\nSummary:",
        temperature=0,
        max_tokens=128,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        #stop=["\n"],
        #n=3,
        #best_of=1,
    )
    summary = response["choices"][0]["text"]
    return summary.replace("\n","").strip()

# Ttile
def title_gpt3(input_text):
    response = openai.Completion.create(
        engine="davinci-instruct-beta-v3",
        prompt=f"Given a transcript, get headline:\n\nTranscript: {input_text}\n\nHeadline:",
        temperature=0,
        max_tokens=32,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        #stop=["\n"],
        #n=3,
        #best_of=1,
    )
    title = response["choices"][0]["text"]
    return title.replace("\n","").strip().strip()

# topics:
def topics_gpt3(input_text):
    response = openai.Completion.create(
        engine="davinci-instruct-beta-v3",
        prompt=f"Given a transcript, get tags:\n\nTranscript: {input_text}\n\nTags:",
        temperature=0,  
        max_tokens=20,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        #stop=["\n"],
        #n=3,
        #best_of=1,
    )
    topics = response["choices"][0]["text"]
    topics = topics[:-1] if topics.endswith(",") else topics
    return topics.replace("\n","").strip()

def sentiment_gpt3(input_text):
    response = openai.Completion.create(
        engine="davinci-instruct-beta-v3",
        prompt=f"Given a transcript, get sentiment:\n\nSentiment: {input_text}\n\nTags:",
        temperature=0,  
        max_tokens=20,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        #stop=["\n"],
        #n=3,
        #best_of=1,
    )
    topics = response["choices"][0]["text"]
    topics = topics[:-1] if topics.endswith(",") else topics
    return topics.replace("\n","").strip()
```

```python
# summarization
def inference(input_text, command, mode="summary"):

    prompt = command[0] + input_text + command[1] 
    try:
        response = openai.Completion.create(
            engine= "davinci-instruct-beta-v3", #self.ENGINE@" #"davinci-instruct-beta-v3", #"ada-instruct-beta",
            prompt=prompt,
            temperature=0,
            max_tokens=130, #self.MAX_LEN_DESCRIPTION+80,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            #stop=["\n"],
            #n=3,
            #best_of=1,
        )
    except:
        # find index of all period in input_text
        periods = [i for i, x in enumerate(input_text) if x == "."]
        period = periods[len(periods)//2]
        input_text = input_text[:period+1]
        
        prompt = command[0] + input_text + command[1]

        response = openai.Completion.create(
            engine= "davinci-instruct-beta-v3", #self.ENGINE", #"davinci-instruct-beta-v3", #"ada-instruct-beta",
            prompt=prompt,
            temperature=0,
            max_tokens=130, #self.MAX_LEN_DESCRIPTION+80,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            #stop=["\n"],
            #n=3,
            #best_of=1,
        )
    
    if mode == "summary":
        summary = response["choices"][0]["text"]
        return summary.replace("\n","").strip()
    elif mode == "title":
        title = response["choices"][0]["text"]
        return title.replace("\n","").strip()
    elif mode == "topics":
        topics = response["choices"][0]["text"]
        topics = topics[:-1] if topics.endswith(",") else topics
        return topics.replace("\n","").strip()
    elif mode == "sentiment":
        sentiment = response["choices"][0]["text"]
        return sentiment.replace("\n","").strip()



command = ["Given a transcript, get summary:\n\nTranscript: ", "\n\nSummary:"]
inference(input_text, command, mode="summary")
```
