```python
lines = f.readlines()
```

```python
lines
```

```python
data = {"add": {}, "mult": {}}
```

```python
for line in lines:
    (_type, n, tp, t, time) = line[:-1].split(" ")
    if n not in data[_type]:
        data[_type][n] = {}
        data[_type][n]["data"] = []
    data[_type][n]["data"].append((tp, t, time))
```

```python
def getMin(_type, n):
    global data
    mini = float(data[_type][str(n)]["data"][0][2])
    tp = data[_type][str(n)]["data"][0][0]
    t = data[_type][str(n)]["data"][0][1]
    for dataline in data[_type][str(n)]["data"]:
        if mini > float(dataline[2]):
            mini = float(dataline[2])
            tp = dataline[0]
            t = dataline[1]
    return (mini, tp, t)
```
