```python
import numpy as np

music = utils.load_audio(MUSIC, duration=5)
speech = utils.load_audio(SPEECH, duration=5)
ms_audio = da.block_mixing_audio(music, speech, overlap=0.5)
ms_spec = preprocessing.get_spectrogram(ms_audio)

mel_original_no_square = preprocessing.get_scaled_mel_bands(ms_spec)
mel_original_squared = preprocessing.get_scaled_mel_bands(ms_spec**2)

music_spec = preprocessing.get_spectrogram(music)
speech_spec = preprocessing.get_spectrogram(speech)
ms_spec_2 = da.block_mixing_spec(music_spec, speech_spec,np.zeros((2, 216)),np.zeros((2, 216)), overlap=0.5)[0]

mel_mixed_no_square = preprocessing.get_scaled_mel_bands(ms_spec_2)
print(np.sum(np.abs(ms_spec - ms_spec_2[:, :-1]) / ms_spec))

music_spec = preprocessing.get_spectrogram(music)**2
speech_spec = preprocessing.get_spectrogram(speech)**2
ms_spec_2, _ = da.block_mixing_spec(music_spec, speech_spec,np.zeros((2, 216)),np.zeros((2, 216)), overlap=0.5)

mel_mixed_squared = preprocessing.get_scaled_mel_bands(ms_spec_2)

print(np.sum(np.abs(ms_spec**2 - ms_spec_2[:, :-1]) / ms_spec**2))
print(np.sum(np.abs(ms_spec - np.sqrt(ms_spec_2[:, :-1])) / ms_spec))



print(np.sum(np.abs(mel_original_no_square - mel_mixed_no_square[:, :-1]) / mel_original_no_square))
print(np.sum(np.abs(mel_original_squared - mel_mixed_squared[:, :-1]) / mel_original_squared))





```
