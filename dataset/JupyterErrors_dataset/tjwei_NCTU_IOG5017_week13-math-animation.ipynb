```python
import sympy
from sympy import symbols, plotting
sympy.init_printing()
```

```python
x, y = symbols('x y')
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plotting.plot_implicit(x**2+ y**2 < 3, color='green')
```

```python
expr_mouth_open = x**2/2+(y+1)**2-.5
expr_cut = abs(y-(-1))+abs(-3-y)-2
expr_mouth = expr_mouth_open**2 + expr_cut**2
plotting.plot_implicit(expr_mouth)
```

```python
expr_face = x**2+ y**2/2-10
expr_eye_left = ((x+1.5)**2+(y-1.7)**2 - .5 )
expr_eye_right = ((x-1.5)**2+(y-1.7)**2 - .5 )

plotting.plot_implicit(expr_eye_right*expr_eye_left*expr_face*expr_mouth*expr_nose)
```

```python
r = x**2+y**2
expr_nose = abs(r-.25)+abs(.5-r)-.25
plotting.plot_implicit(expr_nose)
```

```python
plotting.plot3d(x**2+y**2)
```

```python
plotting.plot_implicit(sympy.And(sympy.Eq(x**2+y**2, 5), x+y < 1))
```

```python
expr_mouth = sympy.And(x**2/2+(y+1)**2-.5, y<-1)
plotting.plot_implicit(expr_mouth)
```

```python
plotting.plot_implicit( (abs(x+3)+abs(x-3)-6)**2 + (x**2+y**2-20)**2 )
```

```python
import matplotlib.pyplot as plt
from IPython.display import display
from ipywidgets import Image
from io import BytesIO
from time import sleep
img = Image()
display(img)
for i in range(1000):
    fig = plotting.plot_implicit( (x-y)**2+i*(x+y)**2-1)
    with BytesIO() as bio:
        fig.save(bio)
        img.value = bio.getvalue()
        sleep(1)
```
