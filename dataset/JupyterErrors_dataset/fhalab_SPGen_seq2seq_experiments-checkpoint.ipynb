```python
class SignalDataset(Dataset):
    
    def __init__(self, h5_file):
        """
        Args:
            h5_file (string): Path to the hdf5 file with the data. 
        """
        self.h5_file = h5py.File(h5_file, 'r')

    def __len__(self):
        return len(self.h5_file['y'])

    def __getitem__(self, idx):
        src = self.h5_file['X'][idx]
        if src[-1, 0] != 1:
            L = len(src)
        else:
            L = np.argwhere(src[:, 0] == 1)[0, 0]
        L = torch.Tensor(np.array([L]))
        src = torch.Tensor(src)
        src = torch.Tensor(self.h5_file['X'][idx])
        tgt = torch.Tensor(self.h5_file['y'][idx])
        return src, tgt, L
```

```python
class AttentionLSTMDecoder(nn.Module):
    
    def __init__(self, in_dim, out_dim, dim, n_layers, dropout=0.1):
        super(AttentionLSTMDecoder, self).__init__()
        self.dim = dim
        self.lstm = nn.LSTM(in_dim, dim, n_layers, dropout=dropout)
        self.dropout = nn.Dropout(p=dropout)
        self.Wa = nn.Parameter(torch.Tensor(dim, dim)) # what exactly is Wa?
        self.tanh = nn.Tanh() # defined activation function
        self.softmax = nn.Softmax(dim=-1) # defined loss function
        self.dense1 = nn.Linear(2 * dim, 2 * dim, bias=False) # linear attention
        self.dense2 = nn.Linear(2 * dim, out_dim)
        self.reset_parameters()
        
    def reset_parameters(self): # purpose of resetting parameters?
        stdv = 1. / np.sqrt(self.Wa.size(1))
        self.Wa.data.uniform_(-stdv, stdv)
        
    def forward(self, tgt, h, c, encoded): # h, c --> hidden state
        """ Decode one step. """
        # tgt is 1 x B x d_out
        # encoded is L x B x D
        decoded, (ht, ct) = self.lstm(tgt, (h, c))
        decoded = self.dropout(decoded)
        # decoded is 1 x B x D
        # Get the attention scores (attention weights)
        # calculating attention weights from encoded, decoded and Wa
        score = torch.matmul(encoded.transpose(0, 1), self.Wa)
        score = torch.bmm(score, decoded.transpose(0, 1).transpose(1, 2))
        att = self.softmax(score.squeeze()).t()
        # att is L x B
        # Element-wise multiplication to weight the encoder outputs
        # Attention weights multiplied by the encoder output vectors to create a
        # weighted combination
        context = att.unsqueeze(-1).repeat(1, 1, self.dim) * encoded
        # This is L x B x D, so sum on dimension 0
        # Now add the weighted encoder outputs
        context = torch.sum(context, dim=0).unsqueeze(0)
        # context is 1 x B x D
        decoded = torch.cat([context, decoded], dim=-1)
        decoded = self.tanh(self.dense1(decoded))
        decoded = self.dropout(decoded)
        decoded = self.dense2(decoded)
        return decoded, ht, ct
```

```python
class Seq2Seq(object):
    
    def __init__(self, in_dim, out_dim, dim, n_layers, dropout=0.1, cuda=False, lr=1e-3, patience=3, factor=0.8):
        self.encoder = nn.LSTM(in_dim, dim, n_layers, dropout=dropout)
        self.decoder = AttentionLSTMDecoder(out_dim, out_dim, dim, n_layers, dropout=dropout)
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.dim = dim
        self.cuda = cuda
        w = torch.ones(out_dim)
        w[0] = 0
        self.loss_func = nn.CrossEntropyLoss(weight=w)
        self.optimizer = optim.RMSprop(list(self.encoder.parameters()) + list(self.decoder.parameters()), lr=lr)
        self.scheduler = optim.lr_scheduler.ReduceLROnPlateau(self.optimizer, factor=factor, patience=patience)
        if cuda:
            if torch.cuda.device_count() > 1:
                self.encoder = nn.DataParallel(self.encoder, dim=1)
                self.decoder = nn.DataParallel(self.decoder, dim=1)
            self.encoder = self.encoder.cuda()
            self.decoder = self.decoder.cuda()
            self.loss_func = self.loss_func.cuda()

        
    def train(self, train_data, val_data, epochs=1, forcing=0.5, chkpt=None, patience=10):
        history = {
            'train_loss': [],
            'train_accu': [],
            'valid_loss': [],
            'valid_accu': []
        }
        for ei in range(epochs):
            print('Epoch %d of %d ' %(ei + 1, epochs))
            start = time.time()
            train_loss, train_accu = self._epoch(train_data, train=True, forcing=forcing)
            history['train_loss'].append(train_loss)
            history['train_accu'].append(train_accu)
            print('\ttime: {elapse:3.3f} s'.format(elapse=(time.time() - start)))
            start = time.time()
            if val_data is not None:
                valid_loss, valid_accu = self._epoch(val_data, train=False, forcing=forcing)
                self.scheduler.step(valid_loss)
                print('\ttime: {elapse:3.3f} s'.format(elapse=(time.time() - start)))
                history['valid_loss'].append(valid_loss)
                history['valid_accu'].append(valid_accu)
                best = np.argmin(history['valid_loss'])
                if chkpt is not None and best == ei:
                    print('Saving checkpoint...')
                    torch.save(self.encoder.state_dict(), chkpt + '_encoder.pkl')
                    torch.save(self.decoder.state_dict(), chkpt + '_decoder.pkl')

                elif (ei - best) > patience:
                    break                
        return history       
        
    def _epoch(self, data, train=True, forcing=0.5):       
        total_loss = 0
        total_acc = 0
        total_len = 0
        if train:
            self.encoder.train()
            self.decoder.train()
        else:
            self.encoder.eval()
            self.decoder.eval()
        for i, batch in enumerate(data):
            src, tgt, L = batch
            L = L.squeeze()
            src = Variable(src)
            tgt = Variable(tgt)
            # Sort by length
            L, inds = torch.sort(L, descending=True)
            src = src[inds]
            src = src.transpose(0, 1)
            tgt = tgt.transpose(0, 1)
            if self.cuda:
                src = src.cuda()
                tgt = tgt.cuda()
            # Pack the src sequences
            src = nn.utils.rnn.pack_padded_sequence(src, L.numpy().astype(int), batch_first=False)
            # forward
            if train:
                self.optimizer.zero_grad()
            pred = self._batch(src, tgt, train=train, forcing=forcing)
            # Get the loss and accuracy
            # Flatten the predictions to B * L x C
            pred = pred.transpose(0, 1).contiguous().view(-1, self.out_dim)
            # Convert to tokens size B * L
            gold = torch.max(tgt[1:].view(-1, self.out_dim), dim=1)[1]
            # The loss expects N x C and N inputs. (B * L = N)
            loss = self.loss_func(pred, gold)
            pred = torch.max(pred, dim=-1)[1].cpu().data.numpy()
            gold = gold.cpu().data.numpy()
            pad = np.where(gold != 0)
            acc = (pred == gold)[pad]
            acc = np.sum(acc) / len(acc)

            if train:
                # backward
                loss.backward()
                # update parameters
                self.optimizer.step()

            # note keeping
            total_loss += loss.data[0] * len(src)
            total_acc += acc * len(src)
            total_len += len(src)
            if train:
                update = '\rTraining batch %d\tloss: %.4f\tacc: %.4f' \
                    %(i + 1, total_loss / total_len, total_acc / total_len)
            else:
                update = '\rValidation batch %d\tloss: %.4f\tacc: %.4f' \
                    %(i + 1, total_loss / total_len, total_acc / total_len)
            print(update, end='')
        return total_loss / total_len, total_acc / total_len
        
    def _batch(self, src, tgt, train=True, forcing=0.5):
        " Get predictions for one batch."
        Lo, Bo, do = tgt.size()
        if do != self.out_dim:
            raise ValueError('Output dimension does not match!')

        
        if train:
            self.encoder.train()
            self.decoder.train()
        else:
            self.encoder.eval()
            self.decoder.eval()
    
        encoded, (hs, cs) = self.encoder(src)
        encoded, _ = torch.nn.utils.rnn.pad_packed_sequence(encoded, batch_first=False, padding_value=0)
        # Initialize decoder hidden state and cell state
        h = hs
        c = cs
        # Run decoder step by step
        decoded = Variable(torch.zeros(Bo, Lo - 1, do))
        if self.cuda:
            decoded = decoded.cuda()
        for i in range(len(tgt) - 1):
            force = True
            if i > 0 and forcing < 1:
                p = torch.rand(1)[0]
                if p > forcing:
                    force = False
            if force:
                tg = tgt[i].unsqueeze(0)
            else:
                # Convert the previous output to 1-hot
                di = decoded[:, i - 1, :]
                maxes = torch.max(di, dim=-1, keepdim=True)[0]
                tg = torch.eq(di, maxes).unsqueeze(0).float()
                if self.cuda:
                    tg = tg.cuda()
            di, h, c = self.decoder(tg, h, c, encoded)
            decoded[:, i, :] = di     
        return decoded
```

```python
# Augmented training
# Mask the attention layer -- no need, already padded with 0s
# Stop decoding after stop token -- easier to 'mask' the loss
# More or less teacher forcing
# teacher forcing decay
# beam search
# accuracy in torch not numpy
# vary hyperparameters

# Vary patience/factor for lr reduction
```

```python
in_dim = 26
n_layers = 2
dec_length = 72
out_dim = 26
bs = 32
d_h = 512
n_layers = 4
dropouts = [0.1, 0.2, 0.3, 0.4]

sd = SignalDataset('../data/train_augmented.hdf5')
train_loader = DataLoader(sd, batch_size=bs, shuffle=True, num_workers=1)
sd = SignalDataset('../data/validate.hdf5')
valid_loader = DataLoader(sd, batch_size=bs, shuffle=False, num_workers=1)


histories = []
for dropout in dropouts:
    fname = 'dropout_' + str(dropout)
    print(fname)
    clf = Seq2Seq(in_dim, out_dim, d_h, n_layers, dropout=dropout, cuda=True, lr=1e-3)
    history = clf.train(train_loader, valid_loader, epochs=1000, 
                        forcing=1.0, patience=10, chkpt='../outputs/' + fname)
    histories.append(history)
```
