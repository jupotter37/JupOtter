```python
import pandas as pd
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from pprint import pprint

from sklearn.model_selection import train_test_split
```

```python
imoveis_df = pd.read_csv("https://raw.githubusercontent.com/nubank/diversidados-curso-ds/master/iniciante/LinReg/dataset_imoveis.csv")


imoveis_df = imoveis_df.drop(columns = "data").sample(1000)
print(imoveis_df.shape)
print(imoveis_df.columns)
imoveis_df.head()
```

```python
target = "preco_dolares"

X_train, X_test, y_train, y_test = train_test_split(imoveis_df.drop(columns = target), imoveis_df[target], test_size=0.33)
```

```python
rf = RandomForestRegressor(random_state = 42)

pprint(rf.get_params())
```

# Iremos otimizar 4 hiperparametros do algortimo Random Forest

### Documentação: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

- n_estimators (int, default=100) - Número de arvores na Random forest - 
- min_samples_split (int or float, default=2) - Minino número de samples para ter uma ramificação 
- max_depth (max_depthint, default=None) - profundidada máxima de uma árvore
- max_features ({“auto”, “sqrt”, “log2”}, int or float, default=”auto”)  - número de máximo de fatures para considerar na hora de fazer a ramificação

```python

n_estimators = [200, 500, 1000]

n_estimators
```

```python
# Produndidade Máxima
max_depth = list(np.linspace(10, 50, num = 3).astype(int))
print(max_depth)


max_depth = list(range(10,31,10))
print(max_depth)


max_depth = list(np.arange(10,31,10))
print(max_depth)
```

```python
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
```

```python
rf = RandomForestRegressor(random_state = 42)

grid = {
"n_estimators": n_estimators, 
"max_depth": max_depth,
"max_features": max_features,   
}
```

Grid Search

```python


grid_search = GridSearchCV(estimator = rf,
                        param_grid = grid,
                        scoring = 'r2', 
                        cv = 3, 
                        verbose = 1)

grid_search.fit(X_train, y_train)
```

```python
pd.DataFrame(grid_search.cv_results_).sort_values("rank_test_score")
```

```python
best_model = grid_search.best_estimator_

best_model.predict(X_train)
```

```python
grid_search.best_params_
```

Random Search

```python
random_search = RandomizedSearchCV(
                        n_iter = 10,
                        estimator = rf,
                        param_distributions = grid,
                        scoring = 'r2', 
                        cv = 5, 
                        verbose = 1,
                        n_jobs = -1)

random_search.fit(X_train, y_train)
```

```python
pd.DataFrame(random_search.cv_results_).sort_values("rank_test_score")
```

```python
best_model = random_search.best_estimator_

best_model.predict(X_train)
```
