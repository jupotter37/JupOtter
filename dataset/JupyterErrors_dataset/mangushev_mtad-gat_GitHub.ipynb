```python
import matplotlib.pyplot as plt

import re
import numpy as np
from pathlib import Path

%matplotlib inline
#1-1
logfile = "../logs/train-machine-1-1-202103272-4X"

def plot_loss(files):
    step = []
    loss = []
    mov_loss=[]
    minusDkl = []
    mov_minusDkl=[]
    log_prob = []
    mov_log_prob=[]    
    mv = 0
    start = True

    p = re.compile('.*\-Dkl = .*')
    p1 = re.compile('.*NaN.*')

    for file in files:
        with open(file) as f:
            lines = f.readlines()
            for line in lines:
                if not p.match(line):
                    continue               
                if p1.match(line):
                    continue
                items = re.split(" +", line)
                if (items[7] != 'loss'):
                    continue
                
                loss_value=float(items[9].split(',')[0])
                if loss_value > 5000:
                    continue                
                loss.append(loss_value)
                
                step.append(int(items[18]))

                minusDkl_value=float(items[6].split(',')[0])
                minusDkl.append(minusDkl_value)
                
                log_prob_value=float(items[15].split(',')[0])
                log_prob.append(log_prob_value) 
                
                if start:
                    mv = loss_value
                else:
                    mv = 0.002*loss_value + 0.998*mov_loss[-1]
                    
                mov_loss.append(mv)

                if start:
                    mv = minusDkl_value
                else:
                    mv = 0.002*minusDkl_value + 0.998*mov_minusDkl[-1]

                mov_minusDkl.append(mv)
                
                if start:
                    mv = log_prob_value
                else:
                    mv = 0.002*log_prob_value + 0.998*mov_log_prob[-1]
                    
                mov_log_prob.append(mv)
                
                start = False
           
    p1, = plt.plot(step[0:], loss[0:], color='magenta',label='loss')
    p2, = plt.plot(step[0:], mov_loss[0:], color='red',label='moving loss')
    #p3, = plt.plot(step[0:], minusDkl[0:], color='green',label='minusDkl')
    #p4, = plt.plot(step[0:], mov_minusDkl[0:], color='black',label='moving minusDkl')
    #p5, = plt.plot(step[0:], log_prob[0:], color='yellow',label='log_prob')
    #p6, = plt.plot(step[0:], mov_log_prob[0:], color='brown',label='moving log_prob')
    
    plt.xlabel("Steps")
    plt.ylabel("Loss")
    #plt.legend(handles=[p1,p2,p3,p4,p5,p6],labels=['loss','moving loss','minusDkl','moving minusDkl','log_prob','moving log_prob'],loc='best')
    #plt.legend(handles=[p1,p3,p5],labels=['loss','minusDkl','log_prob'],loc='best')
    plt.legend(handles=[p1,p2],labels=['loss','moving loss'],loc='best')
    #plt.legend(handles=[p1],labels=['loss'],loc='best')
    plt.xticks(np.arange(0, 30000, 1000))
    plt.yticks(np.arange(0.2, 1.9, step=0.05))
    plt.grid(True, which='both')
    #plt.minorticks_on()
    plt.show()
    
fig = plt.figure(figsize=(24,24))
plot_loss([logfile])
```

```python
import matplotlib.pyplot as plt

import numpy as np

%matplotlib inline

RMS = np.genfromtxt('data/loss/loss_machine-1-1.csvX', delimiter=',', skip_header=0, dtype='float')
label = np.genfromtxt('data/label/machine-1-1.txtX', delimiter=',', skip_header=0, dtype='int')

fig = plt.figure(figsize=(24,12))

p1, = plt.plot(RMS, color='grey',label='RMS')
p2, = plt.plot(label, color='red',label='Anomaly')
p3, = plt.plot(np.repeat(1.075643878, label.shape[0]), color='green', label='98 percentile')
p4, = plt.plot(np.repeat(1.3515916, label.shape[0]), color='magenta', label='POT 0.005 Threshold')
p5, = plt.plot(np.repeat(1.4, label.shape[0]), color='black', label='Best F1 Threshold')

plt.ylabel("Loss")
plt.legend(handles=[p1,p2,p3,p4,p5],labels=['RMS','Anomaly','98 percentile','POT 0.005 Threshold','Best F1 Threshold'],loc='best')
plt.grid(True, which='both')
plt.show()
```
