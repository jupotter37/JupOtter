# Example of optimizing a convex function

```python
from mango.tuner import Tuner
from scipy.stats import uniform
```

```python

param_dict = {
    'x': uniform(-5, 10),
    'y': uniform(-5, 10),
}

    
def objfunc(args_list):
    hyper_evaluated = []
    objective_evaluated = []
    for hyper_par in args_list:
        x = hyper_par['x']
        y = hyper_par['y']
        objective = -(x**2 + y**2)
        objective_evaluated.append(objective)
        hyper_evaluated.append(hyper_par)
    return hyper_evaluated, objective_evaluated
```

```python
conf_Dict = dict()
conf_Dict['batch_size'] = 1
conf_Dict['num_iteration'] = 20
conf_Dict['domain_size'] = 10000
conf_Dict['optimizer']="Random"
tuner_random = Tuner(param_dict, objfunc,conf_Dict)
results_random = tuner_random.maximize()
```

```python
#print('Tuner Configuration')
#print(tuner.getConf())
```

```python
conf_Dict = dict()
conf_Dict['batch_size'] = 1
conf_Dict['initial_random'] = 3
conf_Dict['num_iteration'] = 100
conf_Dict['domain_size'] = 10000
conf_Dict['optimizer']="Bayesian"
tuner_bayesian = Tuner(param_dict, objfunc,conf_Dict)
results_bayesian = tuner_bayesian.maximize()
```

```python
# Random Best Parameters

print('best parameters:',results_random['best_params'])
print('best objective:',results_random['best_objective'])
```

```python
# Bayesian Best Parameters
print('best parameters:',results_bayesian['best_params'])
print('best objective:',results_bayesian['best_objective'])
```

# Actual variation in the objective

```python
Size = 2000

from matplotlib import pyplot as plt
fig = plt.figure(figsize=(30,5))
plt.title('Variation of Objective',fontsize=20)
plt.plot(results_random['objective_values'][:Size],lw=4,label='Random')
plt.plot(results_bayesian['objective_values'][:Size],lw=4,label='BL')
plt.xlabel('Iterations', fontsize=25)
plt.ylabel('objective_values',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(prop={'size': 30})
plt.show()
```
