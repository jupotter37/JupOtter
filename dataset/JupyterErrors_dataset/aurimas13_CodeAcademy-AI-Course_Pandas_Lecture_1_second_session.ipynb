## Working with JSON

```python
# when JSON is a simple dict:
a_dict = {
    'school': 'ABC primary school',
    'location': 'Vilnius',
    'ranking': 1,
}
pd.json_normalize(a_dict)
```

```python
# when a list of dicts
json_list = [
    { 'class': 'Year 1', 'student number': 20, 'name': 'Yellow' },
    { 'class': 'Year 2', 'student number': 25, 'name': 'Blue' },
]
pd.json_normalize(json_list)
```

```python
# when JSON is a nested list
json_obj = {
    'school': 'ABC primary school',
    'location': 'Lithuania',
    'ranking': 1,
    'info': {
        'president': 'Edgaras',
        'contacts': {
            'email': {
                'admission': 'admission@abc.com',
                'general': 'info@abc.com'
            },
            'tel': '123456789',
        }
    },
    'students': [
        { 'name': 'Agne' },
        { 'name': 'Viaceslav' },
        { 'name': 'Antanas'}
    ],
}
pd.json_normalize(json_obj)
```

## Iterating through Groups

```python
nba = pd.read_csv("Data/nba.csv")
positions = nba.groupby('Position')
```

```python
nba_highest_salaries = pd.DataFrame(columns = nba.columns)
nba_highest_salaries
```

```python
for position, data in positions:
    highest_salary_in_nba = data.nlargest(1, 'Salary')
    nba_highest_salaries = nba_highest_salaries.append(highest_salary_in_nba)
nba_highest_salaries
```

```python
nba.groupby('Salary').filter(lambda x: x['Salary'].count() > 12)
```

```python
nba.groupby('Salary').filter(lambda x: x['Salary'].count() > 12)['Salary'].unique()
```

```python
nba['Total Salary Per Position in a league'] = positions['Salary'].transform('sum')
nba.head()
```

```python
nba['Percentage Salary of a Player per position in a league'] = positions['Salary'].fillna(0).transform(lambda x: x/x.sum() * 100)
nba.head()
```

```python
nba = pd.read_csv("Data/nba.csv")
teams = nba.groupby('Team')
nba_team = pd.DataFrame(columns = nba.columns)
nba_team
```

```python
for team, data in teams:
    higest_salary_in_a_team = data.nlargest(1, 'Salary')
    nba_team = nba_team.append(higest_salary_in_a_team)
nba_team.head()
```
