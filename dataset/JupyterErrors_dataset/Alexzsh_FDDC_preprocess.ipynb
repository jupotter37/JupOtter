```python
'''
for llj
'''
import pandas as pd
import numpy as np
import tqdm
def strQ2B(ustring):
    ss = []
    for s in ustring:
        rstring = ""
        for uchar in str(s):
            inside_code = ord(uchar)
            if inside_code == 12288:  # 全角空格直接转换
                inside_code = 32
            elif (inside_code >= 65281 and inside_code <= 65374):  # 全角字符（除空格）根据关系转化
                inside_code -= 65248
            rstring += chr(inside_code)
        ss.append(rstring)
    return ss
df = pd.read_csv('cnki.csv')
df['content']=df['title']+'.'+df['abstract']
df['content'].astype(str)
df['content']=strQ2B(df['content'])
_content=df['content'].values.tolist()
_keywords=df['keywords'].str.split(';')
content=_content[int(len(_content)*0.9):]
keywords=_keywords[int(len(_keywords)*0.9):]
for i,keyword in tqdm.tqdm(enumerate(keywords)):
    sentence=content[i]
    label=[]
    for j,word in enumerate(keyword):
        if word in sentence:
            label.append((sentence.find(word),len(word)))
    sorted(label,key=lambda r:r[0])
    tag=[]
    if len(label)==0: continue
    tag.extend('O'*label[0][0])
    for j,l in enumerate(label):
        tag.append('B-KEY')
        tag.extend(['I-KEY']*(l[1]-1))
        if j !=len(label)-1:
            tag.extend('O'*(label[j+1][0]-l[0]-l[1]))
    tag.extend('O'*(len(sentence)-label[-1][0]-label[-1][1]))
    with open('../NER_IDCNN_CRF/data/cnki.dev','a',encoding='utf-8') as fw:
        for j,s in enumerate(sentence):
            fw.write(sentence[j]+' '+tag[j]+'\n')
        fw.write('\n')
```
