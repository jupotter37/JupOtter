# Debugging Notebooks 


# Naive Way - print


```python
import random
def find_max (values):
    max = 0
    print(f"Initial max is {max}")
    for val in values:
        if val > max:
            max = val
    return max

sample = random.sample(range(100), 10)
find_max(sample)
```

### Advantages
* Easy 
* No installation required

### Disadvantages:
* Hard to pinpoint error prown locations
* Can be spammy

# Classical Way - PDB


### PDB is used in the following ways:
* using "breakpoint()" since python 3.7.
* using "from IPython.core.debugger import set_trace;set_trace()" for python notebooks.
* using "import pdb; pdb.set_trace()".

```python
import random
def find_max (values):
    max = 0
    import pdb; pdb.set_trace()
    for val in values:
        if val > max:
            max = val
    return max

sample = random.sample(range(100), 10)
find_max(sample[:-3])
```

    * Less bugs

* Very scary

# Better Way Jupyter Notebooks - pixie_debugger

```python
import contextlib

with contextlib.redirect_stdout(None):
    import pixiedust
```

```python
%%pixie_debugger
import random
def find_max (values):
    max = 0
    for val in values:
        if val > max:
            max = val
    return max
find_max(random.sample(range(100), 10))
```

### Advantages
* Easy 
* Dynamic

### Disadvantages:
* Additional installation required
* Not mature:
    * Documentation
    * Supported evaluation
    * Working on Jupyter notebooks but not in jupyterlab
