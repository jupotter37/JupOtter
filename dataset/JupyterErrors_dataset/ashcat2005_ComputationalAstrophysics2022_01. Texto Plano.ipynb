## 1. Lectura de archivos de texto plano

Utilizando [numpy](https://numpy.org), es posible leer desde un archivo de texto plano y cargarlos directamente a un arreglo. 

```python
import numpy as np

data = np.loadtxt('example01.txt', comments='#')
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.plot(data[:,0],data[:,1], color='crimson')
plt.show()
```

Cuando los datos en el archivo est√°n separados por un simbolo diferente al espacio, debe escribirse con el argumento `delimiter`.

```python
import numpy as np

data = np.loadtxt('example02.txt', comments='#')
```

```python
import numpy as np

data = np.loadtxt('example02.txt', comments='#', delimiter=',')
data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.plot(data[:,0],data[:,1], color='crimson')
plt.show()
```

##¬†02. Escritura de archivos de texto plano

De nuevo, la librer√≠a [numpy](https://numpy.org) tiene una funci√≥n sencilla para la escritura de archivos de texto plano.

```python
def f(x):
    return np.exp(-x**2/5)

wdata = np.zeros([2,100])

wdata[0,:] = np.linspace(-10,10,100)
wdata[1] = f(wdata[0])


np.savetxt('out_file1.txt', wdata, fmt='%5.5f', header='This is the header of the file')
```

```python
outdata = np.loadtxt('out_file1.txt', comments='#')


plt.figure(figsize=(10,7))
plt.plot(outdata[0],outdata[1], color='crimson')
plt.show()
```

```python
np.savetxt('out_file2.data', wdata, delimiter = ',', 
           fmt='%5.5f', header='This is the header of the file')

outdata2= np.loadtxt('out_file2.data', comments='#', delimiter=',')


plt.figure(figsize=(10,7))
plt.plot(outdata2[0],outdata2[1], color='crimson')
plt.show()
```
