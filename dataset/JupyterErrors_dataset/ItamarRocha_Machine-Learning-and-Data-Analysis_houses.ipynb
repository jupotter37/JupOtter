```python
y = dataset1.iloc[:,8].values
X = dataset1.drop(columns = ['median_house_value']).values
print(y)
```

## We are going to scale the data since we are using distance methods such as svm and knn    

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = y.reshape(-1,1)
y = scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

using pca to get good resume

## Lets implement all algorithms that we know 

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

### 1. SUPORT VECTOR MACHINE

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf',gamma = 1)
regressor.fit(X_train,y_train)
```

```python
accuracies = cross_val_score(estimator = regressor, X = X, y = y, cv = 10,n_jobs=-1)
```

```python
accuracies.mean()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'kernel': ['rbf'],'gamma': [0,1], 'C': [0.5,1]},
              {'kernel': ['sigmoid'],'gamma': [0,1], 'C': [0.5,1]}]
grid = GridSearchCV(estimator = regressor,
                    param_grid= parameters,
                    cv = 10,
                    n_jobs=-1)


grid = grid.fit(X_train,y_train)
best_accuracy = grid.best_score_
best_parameters = grid.best_params_
print(best_parameters)
print(best_accuracy)
```

### 2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

```python
accuracies = cross_val_score(estimator = regressor, X = X, y = y, cv = 10,n_jobs=-1)
```

```python
accuracies.mean()
```

### 3. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators= 10)
regressor.fit(X_train, y_train)
```

```python
accuracies = cross_val_score(estimator = regressor, X = X, y = y, cv = 10,n_jobs=-1)
```

```python
accuracies.mean()
```

###  4. KNN

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors= 3)
regressor.fit(X_train,y_train)
```

```python
accuracies = cross_val_score(estimator = regressor, X = X, y = y, cv = 10,n_jobs=-1)
```

```python
accuracies.mean()
```

### 5. XGBOOST

```python
from xgboost import XGBClassifier #fast and you dont need feature scalling
regressor = XGBClassifier()
regressor.fit(X_train, y_train.ravel())
```

```python
accuracies = cross_val_score(estimator = regressor, X = X, y = y, cv = 10,n_jobs=-1)
```
