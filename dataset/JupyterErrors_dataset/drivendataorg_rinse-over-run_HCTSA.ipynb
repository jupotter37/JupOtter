```
import pandas as pd
import numpy as np
import pyopy
from pyopy.hctsa import hctsa
import time
```

```
def extract_feature(ts, operation):
    # Start hctsa Engine usying pyopy
    hctsa.prepare(engine='octave')
    # Try extracting the result
    result = hctsa.operations.__dict__[operation.strip()](ts)
```

```
_ = hctsa.prepare(engine='octave')
```

```
operators = hctsa.catalog.operations_dict.keys()
```

```
x = np.sin(np.arange(-np.pi, np.pi, 1000)) + np.random.RandomState(42).randn(1000)
fast = []
for op in operators:
  try:
    start = time.time()
    extract_feature(x, op)
    total_time = time.time() - start
    print('SUCCESS:', op, total_time)
    if total_time < 3:
      fast.append(op)
  except:
    print('FAIL:', op)
print(fast)
```
