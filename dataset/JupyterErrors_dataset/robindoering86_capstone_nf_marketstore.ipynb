```python
from atomm.DataManager.main import MSDataManager
import pandas as pd
# 
import pymarketstore as pymkts
import numpy as np
```

```python
## Copying thetolisttire dataset to the database

df2 = pd.read_csv('/Users/robin/Documents/nf-ds/capstone_nf/data/raw/historical_stock_prices.csv')
dm = MSDataManager()
symbols_in_db = dm.ListSymbols()

cli = pymkts.Client('http://atomm.de:5993/rpc')
cnt = 1
symb_list = df2['ticker'].unique()

total = len([x for x in symb_list if x not in symbols_in_db])

for symb in symb_list:
    if symb not in symbols_in_db:
        df = df2[df2['ticker'] == symb].copy()
        df.set_index('date', inplace=True,)
        df.index = [pd.Timestamp(x) for x in df.index]
        df_kg1 = df.copy()
        data = np.array([(df_kg1.index[i].value / 10**9, df_kg1['open'][i], df_kg1['high'][i], df_kg1['low'][i], df_kg1['close'][i], df_kg1['volume'][i]) for i in range(len(df_kg1))],
                             dtype=[('Epoch', 'i8'), ('Open', 'f4'), ('High', 'f4'), ('Low', 'f4'), ('Close', 'f4'), ('Volume', 'i4')])
        cli.write(data, symb+'/1D/OHLCV')
        print(f'[{cnt}/{total}] Stored historical price data for symbol {symb}\n for period: {df_kg1.index[0]} - {df_kg1.index[-1]}\n\n')
        cnt += 1
```
