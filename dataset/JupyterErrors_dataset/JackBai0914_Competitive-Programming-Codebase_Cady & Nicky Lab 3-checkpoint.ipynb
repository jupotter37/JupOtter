## Collider Kinematics

We begin by introducing the CMS detector and discussing the kinematics of the collision events.  Here is a picture of the CMS detector:

CMSdetector.png

The proton beams (inside the orange tubes) come in from both sides of the image above and collide in the central region.  An assortment of particles can be produced in these collisions and we need ways to detect them. The inner trackers detect charged particles that are produced in the interaction.  Next comes the electromagnetic calorimeter, which measures the energies and momenta of electrons and photons.  The hadron calorimeter is designed to accurately measure the energies and momenta of quark and gluon bound states, like protons and neutrons.  The muon chambers on the outside of the detector record the properties of muons, which are charged particles that are similar to electrons, but heavier.  Taken together, the CMS detector is truly massive.  The little human in the image above is drawn to scale.  

Moving forward, we we will take as our convention that the proton beams come in 

along the $z$ axes, as shown below.

axes.png

The interaction vertex is located at the origin of the coordinate axes above.  It turns out to be convenient to work with a set of variables that are the same in the lab frame as well as the center-of-mass frame of the collision.  This means that we want to find variables that are invariant to transformations between frames, even when the frames are moving at relativistic velocities (i.e., close to the speed of light) relative to each other.  The fancy way of saying this is that the variables are invariant to "Lorentz boosts'' between frames.  (We will define Lorentz transformations in the special relativity mini-course later this semester...so stay tuned!)

The transverse momenta $p_x$ and $p_y$ are Lorentz boost invariant.  The scalar transverse momentum is defined as 

$$
p_T = \sqrt{p_x^2 + p_y^2} \,.
$$

The azimuthal angle 

$$
\phi \equiv \tan^{-1} \frac{p_x}{p_y}
$$

is also Lorentz boost invariant.  Another boost-invariant quantity is the pseudorapidity, $\eta$, which is defined as 

$$
\eta \equiv \ln \cot \frac{\theta}{2} \, ,
$$

where $\cos\theta = p_z/|\vec{p}|$.

It is possible for more than one collision to occur in the detector at the same time because each proton beam actually consists of "bunches" of protons.  This obviously complicates the analysis and we need to make sure that the electrons and muons that we select come from the *same* primary interaction vertex. (For exmaple, it would be problematic if we selected a pair of electrons that came from a different collision event than the pair of muons that we selected.). To guarantee this, we must require the following:

1. |Muon_dxy| < 0.5 and |Electron_dxy| < 0.5
2. |Muon_dz| < 1.0 and |Electron_dz| < 1.0 
3. sip3d < 4 for all muons and electrons

The first two items are conditions on the impact parameter of the muon/electron relative to the primary vertex.  The third item depends on the quantity sip3d, which is defined as 

sqrt(Muon_dxy^2 + Muon_dz^2)/sqrt(Muon_dxyErr^2 + Muon_dzErr^2)

and similarly for the electron. 

Define a function that evaluates sip3d for either a muon or an electron.  Then, apply the primary vertex selection criteria given above, in addition to the isolation and kinematic cuts that you already applied.  How many events are left after you apply these selections?  Convince the reader that cuts are being applied properly.

*Hint*: Some useful functions for this exercise are [np.power()](https://numpy.org/doc/stable/reference/generated/numpy.power.html) and [np.divide()](https://numpy.org/doc/stable/reference/generated/numpy.divide.html).



````python
def get_sip3d(dxy, dz, dxyErr, dzErr):
    sip3d = math.sqrt(dxy**2 + dz**2)/math.sqrt(dxyErr**2 + dzErr**2)
    return sip3d
````

````python
mu_sip3d_array = []
for i in range(len(filter2_dict["Muon_dxy"])):
    mu_sip3d_event = []
    for j in range (len(filter2_dict["Muon_dxy"][i])):
        dxy = filter2_dict["Muon_dxy"][i][j]
        dz = filter2_dict["Muon_dz"][i][j]
        dxyErr = filter2_dict["Muon_dxyErr"][i][j]
        dzErr = filter2_dict["Muon_dzErr"][i][j]
        sip3d = get_sip3d(dxy, dz, dxyErr, dzErr)
        mu_sip3d_event.append(sip3d)
    mu_sip3d_array.append(mu_sip3d_event)

e_sip3d_array = []
for i in range(len(filter2_dict["Electron_dxy"])):
    e_sip3d_event = []
    for j in range (len(filter2_dict["Electron_dxy"][i])):
        dxy = filter2_dict["Electron_dxy"][i][j]
        dz = filter2_dict["Electron_dz"][i][j]
        dxyErr = filter2_dict["Electron_dxyErr"][i][j]
        dzErr = filter2_dict["Electron_dzErr"][i][j]
        sip3d = get_sip3d(dxy, dz, dxyErr, dzErr)
        e_sip3d_event.append(sip3d)
    e_sip3d_array.append(e_sip3d_event)
    
filter2_dict["Muon_sip3d"] = np.array(mu_sip3d_array)
filter2_dict["Electron_sip3d"] = np.array(e_sip3d_array)

mu_dxy_condition = np.array([all([x>-0.5 and x<0.5 for x in array]) for array in filter2_dict['Muon_dxy']]) 
e_dxy_condition = np.array([all([x>-0.5 and x<0.5 for x in array]) for array in filter2_dict['Electron_dxy']]) 
mu_dz_condition = np.array([all([x>-1 and x<1 for x in array]) for array in filter2_dict['Muon_dz']]) 
e_dz_condition = np.array([all([x>-1 and x<1 for x in array]) for array in filter2_dict['Electron_dz']]) 
mu_sip3d_condition = np.array([all([x<4 for x in array]) for array in filter2_dict['Muon_sip3d']]) 
e_sip3d_condition = np.array([all([x<4 for x in array]) for array in filter2_dict['Electron_sip3d']]) 

all_satisfied = np.array(np.all([mu_dxy_condition, e_dxy_condition, mu_dz_condition, e_dxy_condition, mu_sip3d_condition, e_sip3d_condition], axis = 0))

filter3_dict = {}
for key, value in filter2_dict.items():  
  filter3_dict[key] = value[all_satisfied]

print("Number of events left in sample: " + str(len(filter3_dict["Muon_charge"])))
````

We define the invariant mass between two particles to be 

$$
m_{2L} = \sqrt{(E_1 + E_2)^2 - (\vec{p_1} + \vec{p_2})^2} \, ,
$$

where $E_1$ and $\vec{p_1}$ are the energy and momenta of particle 1 and where $E_2$ and $\vec{p_2}$ are the energy and momenta of particle 2.  As we will discuss in the special relativity mini-course, the invariant mass is a special mass that is invariant under Lorentz transformations between frames.  

The invariant mass of two particles that are produced from the decay of a heavier particle will be related to the mass of the heavier parent particle.  When the parent is actually produced in the detector (as opposed to being what we call "virtual" -- yes, that's a thing...but explaining this would take us too far off course :)), then the invariant mass will closely reproduce the mass of the parent.    

When searching for the Higgs, we require that one pair of leptons has an invariant mass that is quite close to the mass of the $Z$-boson (~90 GeV), while the invariant mass of the other pair is not too far off.  Specifically, we require that

1. the pair with invariant mass closest to the $Z$ mass has $40 < m_{2L} < 120$ GeV
2. the other pair has an invariant mass $12 < m_{2L} < 120$ GeV

Apply these selection criteria to your simulation data sample, in addition to the isolation, kinematic, and primary vertex selections described above.  We recommend that you start off by writing a function that evaluates the invariant mass for a pair of electrons/muons in the sample.  To do so, you can use the following relations for energy and momenta:

$E = p_T \cosh\eta$

$p_x = p_T \sin\phi$

$p_y = p_T \cos\phi$

$p_z = p_T \sinh\eta$

````python
def get_invariant_mass_2L(pt1, eta1, phi1, pt2, eta2, phi2): 
    E1 = pt1*math.cosh(eta1)
    E2 = pt2*math.cosh(eta2)
    px1 = pt1 * math.sin(phi1)
    px2 = pt2 * math.sin(phi2)
    py1 = pt1 * math.cos(phi1)
    py2 = pt2 * math.cos(phi2)
    pz1 = pt1 * math.sinh(eta1)
    pz2 = pt2 * math.sinh(eta2)
    sum_p1p2 = math.sqrt((px1+px2)**2+(py1+py2)**2+ (pz1+pz2)**2)
    invariant_mass = math.sqrt((E1 + E2)**2 - sum_p1p2**2)
    return invariant_mass

invariant_mass_array = []
for i in range(len(filter3_dict["Muon_pt"])):
    invariant_mass_event = []
    pt1 = filter3_dict["Muon_pt"][i][0]
    eta1 = filter3_dict["Muon_eta"][i][0]
    phi1 = filter3_dict["Muon_phi"][i][0]
    pt2 = filter3_dict["Muon_pt"][i][1]
    eta2 = filter3_dict["Muon_eta"][i][1]
    phi2 = filter3_dict["Muon_phi"][i][1]
    mu_invariant_mass = get_invariant_mass_2L(pt1, eta1, phi1, pt2, eta2, phi2)
    invariant_mass_event.append(mu_invariant_mass)
    pt1 = filter3_dict["Electron_pt"][i][0]
    eta1 = filter3_dict["Electron_eta"][i][0]
    phi1 = filter3_dict["Electron_phi"][i][0]
    pt2 = filter3_dict["Electron_pt"][i][1]
    eta2 = filter3_dict["Electron_eta"][i][1]
    phi2 = filter3_dict["Electron_phi"][i][1]
    e_invariant_mass = get_invariant_mass_2L(pt1, eta1, phi1, pt2, eta2, phi2)
    invariant_mass_event.append(e_invariant_mass)
    invariant_mass_array.append(invariant_mass_event)


filter3_dict["Invariant_masses"] = np.array(invariant_mass_array)

def is_quite_close (invariant_mass):
    if 40 < invariant_mass and 120 > invariant_mass:
        return True
    else:
        return False

def not_far_off(invariant_mass):
    if 12 < invariant_mass and 120 > invariant_mass:
        return True
    else:
        return False

def meets_mass_condition(invariant_mass_array):
    if not_far_off(invariant_mass_array[0]) and not_far_off(invariant_mass_array[1]):
            if is_quite_close(invariant_mass_array[0]) or is_quite_close(invariant_mass_array[1]):
                return True
    else:
        return False
        
invariant_mass_condition = np.array([meets_mass_condition(array)==True for array in filter3_dict['Invariant_masses']]) 

filter4_dict = {}
for key, value in filter3_dict.items():  
    filter4_dict[key] = value[invariant_mass_condition]

print("Number of events left in sample: " + str(len(filter4_dict["Muon_charge"])))
plt.hexbin(filter4_dict["Invariant_masses"][:,0], filter4_dict["Invariant_masses"][:,1], cmap = "BuGn")
plt.title("2D Hexagonal Binning Plot of Electron Invariant Mass versus Muon Invariant Mass")
plt.xlabel("Muon Invariant Mass")
plt.ylabel("Electron Invariant Mass")
````

The histogram function, as written above, reweights each event.  That is, the sum of events in each histogram bin is instead the sum of the weights of each event.  This needs to be done to rescale the "quantity" of simulation data to be consistent with the "quantity" of actual data collected by CERN.  This will allow us to make an apples-to-apples comparison of the simulation to the real data later on in the lab.  Additionally, notice that ```stacked=True``` is set in ```plt.hist()```.  This will be important in the next section because it will allow you to stack the histograms for the simulated background and signal.

````python
def get_invariant_mass_4L(pt1, eta1, phi1, pt2, eta2, phi2, pt3, eta3, phi3, pt4, eta4, phi4): 
    E1 = pt1*math.cosh(eta1)
    E2 = pt2*math.cosh(eta2)
    E3 = pt3*math.cosh(eta3)
    E4 = pt4*math.cosh(eta4)
    px1 = pt1 * math.sin(phi1)
    px2 = pt2 * math.sin(phi2)
    px3 = pt3 * math.sin(phi3)
    px4 = pt4 * math.sin(phi4)
    py1 = pt1 * math.cos(phi1)
    py2 = pt2 * math.cos(phi2)
    py3 = pt3 * math.cos(phi3)
    py4 = pt4 * math.cos(phi4)
    pz1 = pt1 * math.sinh(eta1)
    pz2 = pt2 * math.sinh(eta2)  
    pz3 = pt3 * math.sinh(eta3)
    pz4 = pt4 * math.sinh(eta4)
    sum_p1p2p3p4 = math.sqrt((px1+px2+px3+px4)**2+(py1+py2+py3+py4)**2+ (pz1+pz2+pz3+pz4)**2)
    invariant_mass = math.sqrt((E1 + E2 + E3 + E4)**2 - sum_p1p2p3p4**2)
    return invariant_mass

invariant_mass_array = []
for i in range(len(filter4_dict["Muon_pt"])):
    invariant_mass_event = []
    pt1 = filter4_dict["Muon_pt"][i][0]
    eta1 = filter4_dict["Muon_eta"][i][0]
    phi1 = filter4_dict["Muon_phi"][i][0]
    pt2 = filter4_dict["Muon_pt"][i][1]
    eta2 = filter4_dict["Muon_eta"][i][1]
    phi2 = filter4_dict["Muon_phi"][i][1]
    pt3 = filter4_dict["Electron_pt"][i][0]
    eta3 = filter4_dict["Electron_eta"][i][0]
    phi3 = filter4_dict["Electron_phi"][i][0]
    pt4 = filter4_dict["Electron_pt"][i][1]
    eta4 = filter4_dict["Electron_eta"][i][1]
    phi4 = filter4_dict["Electron_phi"][i][1]
    invariant_mass = get_invariant_mass_4L(pt1, eta1, phi1, pt2, eta2, phi2, pt3, eta3, phi3, pt4, eta4, phi4)
    invariant_mass_array.append(invariant_mass)

SMHiggs_m4L = np.array(invariant_mass_array)

# Number of bins for all histograms
nbins = 25

# Weights
luminosity = 11580.0  # Integrated luminosity of the data samples
xsec_SMHiggsToZZTo4L = 0.0065  # H->4l: Standard Model cross-section
nevt_SMHiggsToZZTo4L = 299973.0  # H->4l: Number of simulated events
weight_sig_4mu = luminosity * xsec_SMHiggsToZZTo4L / nevt_SMHiggsToZZTo4L

#Create the histogram
plt.hist([SMHiggs_m4L], bins = nbins, range=(70, 180), histtype = 'step', stacked=True, weights = [np.full(len(SMHiggs_m4L), weight_sig_4mu)], label = ['Signal'])
plt.xlabel(r'$m_{4L}$ [GeV]', fontsize= 12)
plt.ylabel('Events', fontsize = 12)
plt.legend(frameon=False)
plt.show()
````
