```python
def return_z_scores(df, mode='log'):
    if mode=='log': return (np.log(d).diff(1) - np.log(d).diff(1).mean()) / np.log(d).diff(1).std(ddof=0)
    else: return (d.pct_change().diff(1) - d.pct_change().diff(1).mean()) / d.pct_change().diff(1).std(ddof=0) 
rz = abs(return_z_scores(d))
rz.plot(subplots=True, layout=(4, 4), figsize=(20, 15));
```

Fitting pareto on `pct_change` with abs(zscores) > 1

```python
rz_pct = rz_pct.dropna()
rz = rz.dropna()
```

```python
params = []
for c in rz_pct.columns:
    params.append((c, pareto.fit(rz_pct[c][rz_pct[c]>1], 4, loc=0, scale=1)))
```

```python
params
```

```python
currs = [c for c in rz_pct.columns]
fig, ax = plt.subplots(2, 4, figsize=(20, 10))
for i, c in enumerate(rz_pct.columns):
    r = rz_pct[c][rz_pct[c]>1].dropna()
    b, loc, scale =params[i][1][0], params[i][1][1], params[i][1][2]
    x = np.linspace(pareto.ppf(0.01, b, loc=loc, scale=scale), \
                  pareto.ppf(0.99, b, loc=loc, scale=scale), 100)
    ax[i//4][i%4].plot(x, pareto.pdf(x, b, loc=loc, scale=scale), \
           'r-', lw=5, alpha=0.6, label='pareto pdf')
    ax[i//4][i%4].hist(r, density=True, histtype='stepfilled', alpha=0.4, lw=5, bins=1000)
    ax[i//4][i%4].legend(['Pareto fit on {}'.format(currs[i]),'observations'])
```

## Appendix: Cluster Analysis

```python
from bokeh.io import output_file, output_notebook, show
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar, LogTicker, DataRange1d
from bokeh.transform import linear_cmap
from bokeh.palettes import Plasma
from bokeh.plotting import figure
```

```python
output_notebook()
output_file('NDF Spot tSNE')
```

```python
def calc_realized_variance(dataframe, window=30):
    log_returns = np.log(dataframe).diff(1)
    log_returns = log_returns**2
    variance = log_returns.rolling(window).sum()

    return variance
```

```python
rvar = calc_realized_variance(d).mean(axis=1).values
means = d.mean(axis=1)
```

```python
rg = range(len(d.index))

mapper = LinearColorMapper(palette='Magma256' , low=min(rg), high=max(rg))

x = pd.to_datetime(d.index)
source = ColumnDataSource(data=dict(x=x, y1=embeddings[:,0], 
                          y2=embeddings[:,1], y3=means,
                          rg=rg, rvar=rvar))

TOOLS = "box_select, lasso_select, help"


s1 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='tSNE of Spot Rate Vector, NDF FX Pairs')
s1.circle('y1', 'y2', size=4, alpha=0.5, source=source, color={'field': 'rg', 'transform': mapper})

s3 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='Mean Price of NDF FX Pairs',
           x_axis_type="datetime")
s3.line('x', 'y3', alpha=0.5, line_width=2, source=source)

s2 = figure(tools=TOOLS, plot_width=850, plot_height=850, title='Mean Realized Variance of NDF FX Pairs',
           x_axis_type="datetime")
s2.line('x', 'rvar', alpha=0.5, line_width=2, source=source)

p = gridplot([[s1, s2, s3]], toolbar_location=None)

color_bar = ColorBar(color_mapper=mapper, ticker=LogTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))
s1.add_layout(color_bar, 'right')

# show the results
show(p)
```

## Appendix III: Clustering with Rolling Window Normalization

```python
def as_normed_npy(data, n_cols, window=None):
    if window:
        periods = data.iloc[:, :n_cols].rolling(window)
        data_avg = periods.mean().values
        data_std = periods.std().values
    else:
        data_avg = data.iloc[:, :n_cols].mean().values
        data_std = data.iloc[:, :n_cols].std().values

    return (data - data_avg) / data_std
```

```python
d_roll_norm = as_normed_npy(d, 8, 360).dropna()
```

## Appendix V: Determining optimal window for rolling training and validation

```python
c = d.columns
```

```python
pca_info = []
for window in range(10, 360, 10):
    pre_emb = (d - d.rolling(window).mean())/d.rolling(window).std()
    pca = PCA(8)
    emb = pca.fit_transform(pre_emb.dropna())
    c = pca.explained_variance_ratio_
    rel_components = sum(pca.explained_variance_ratio_)**2/sum(pca.explained_variance_ratio_**2)
    pca_info.append((window, rel_components))
```

```python
plt.plot([el[0] for el in pca_info], [el[1] for el in pca_info], color='black', alpha=0.8)
plt.xlabel('Normalization window size')
plt.ylabel('Number of PCA factors')
plt.title('PCA factors for NDF pair spots')
```
