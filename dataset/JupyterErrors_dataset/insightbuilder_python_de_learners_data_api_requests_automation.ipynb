```python
#introducing post request 
# visit https://pyrite-ethereal-soccer.glitch.me/docs for info
pyrite_url = 'https://pyrite-ethereal-soccer.glitch.me/user/v1/show_users'
```

```python
# Making a POST request
url_post = "https://pyrite-ethereal-soccer.glitch.me/user/create_user"
payload = {"email_id": "my_python@pythonuuniverse.com",
           "pass_phrase": "letsmakebetter"}
```

```python
post_payload = requests.post(url=url_post,json=payload)
post_payload.json()
```

```python
requests.get(url=pyrite_url).json()
```

Object serialization is the process of converting an object into byte stream. This byte stream is stored in file-like object. After that it can be transmitted via sockets etc. 

Deserialization is the process of reconstructing the object from the byte stream.

```python
type(post_payload.json())
```

```python
serialized_payload = json.dumps(post_payload.json())
type(serialized_payload)
```

```python
serialized_payload
```

```python
with open('serialized_video_tutorial.txt','w') as js:
    json.dump(post_payload.json(),fp=js)
```
