```python
import pandas as pd
from slugify import slugify
```

```python
column_names = [
"geonameid",
"name",
"asciiname",
"alternatenames",
"latitude",
"longitude",
"feature class",
"feature code",
"country code",
"cc2",
"admin1 code",
"admin2 code",
"admin3 code",
"admin4 code",
"population",
"elevation",
"dem",
"timezone",
"modification date"
]
```

```python
df = pd.read_csv('../files_original/original.geonames.cities500.txt', sep="\t", header=None, names=column_names)
df
```

```python
# filter out out of scope countries
df['region'] = df.timezone.apply(lambda x: x.split('/')[0])
df['region'].value_counts().head(10)
df = df[~df['region'].isin(['Asia', 'Australia', 'America', 'Pacific', 'Indian'])]
df = df[~df['timezone'].isin(['Africa/Lagos', 'Africa/Johannesburg', 'Africa/Dar_es_Salaam'])]
# df[df['region'] != 'Europe']['timezone'].value_counts().head(30)
df
```

```python
def get_names(row):
    names = [row['name']] 
    names += (row['alternatenames'].split(',') if isinstance(row['alternatenames'], str) else [])
    return names
df['all_names'] = df.apply(get_names, axis=1)
```

```python
df_all = df.explode('all_names')
df_all
```

```python
df_all = df_all[['geonameid', 'name', 'all_names', 'latitude', 'longitude', 'country code', 'population']]
df_all.rename(columns={'all_names': 'name_alt', 'name': 'name_main'}, inplace=True)
df_all['name_alt_slug'] = df_all.name_alt.apply(lambda x: slugify(x))
df_all
```

```python
df_by_slug = df_all.set_index('name_alt_slug')
df_by_slug
```

```python
df_scraped = pd.read_json('../files_original/original.greenferries.scraped_ship_routes.json')
df_scraped.rename(columns={'cityA': 'city_a', 'cityB': 'city_b', 'company_name': 'company', 'shipName': 'ship'}, inplace=True)
df_scraped
```

```python
all_cities = set(list(df_scraped.city_a) + list(df_scraped.city_b))
```

```python
df_scraped_cities = pd.DataFrame(pd.Series(list(all_cities)))
df_scraped_cities.columns = ['scraped_name']
df_scraped_cities['scraped_slug'] = df_scraped_cities.scraped_name.apply(slugify)
df_scraped_cities.set_index('scraped_slug', inplace=True)
df_scraped_cities['scraped_slug'] = df_scraped_cities.index
df_scraped_cities = df_scraped_cities.join(df_by_slug)
df_misses = df_scraped_cities[df_scraped_cities.geonameid.isnull()]
df_scraped_cities_matched = df_scraped_cities[~df_scraped_cities.geonameid.isnull()]
df_scraped_cities_matched = df_scraped_cities_matched.groupby('scraped_slug').first()
df_scraped_cities_matched
```

```python
df_misses
# [print(r) for r in misses.index]
for slug, geonameid in manual_fixes.items():
    df_misses.at[slug, 'geonameid'] = geonameid
df_misses
```

```python
df_misses_matched = df_misses[~df_misses['geonameid'].isnull()]
df_misses_matched['scraped_slug'] = df_misses_matched.index
df_misses_matched = df_misses_matched[['scraped_name', 'scraped_slug', 'geonameid']]
df_misses_matched.set_index('geonameid', inplace=True)
df_misses_fixed = df_misses_matched.join(df_all.set_index('geonameid'))
df_misses_fixed = df_misses_fixed.groupby('geonameid').first()
df_misses_fixed['geonameid'] = df_misses_fixed.index
df_misses_fixed.set_index('scraped_slug', inplace=True)
df_misses_fixed = df_misses_fixed[['scraped_name', 'geonameid', 'name_main', 'name_alt', 'latitude', 'longitude', 'country code', 'population']]
df_misses_fixed
```

```python
df_scraped_cities_matched_and_fixed = pd.concat([df_scraped_cities_matched, df_misses_fixed])
df_scraped_cities_matched_and_fixed = df_scraped_cities_matched_and_fixed[~df_scraped_cities_matched_and_fixed['country code'].isnull()]
df_scraped_cities_matched_and_fixed
```

```python
df_scraped['city_a_slug'] = df_scraped.city_a.apply(slugify)
df_scraped['city_b_slug'] = df_scraped.city_b.apply(slugify)
```

```python
def get_city(scraped_city_name):
    slug = slugify(scraped_city_name)
    if slug not in df_scraped_cities_matched_and_fixed.index:
        return {}
    city = df_scraped_cities_matched_and_fixed.loc[slug]
    return city

for suffix in ['a', 'b']:
    df_scraped[f'city_{suffix}_geonameid'] = df_scraped[f'city_{suffix}'].apply(lambda x: get_city(x).get('geonameid'))
    df_scraped[f'city_{suffix}_country_code'] = df_scraped[f'city_{suffix}'].apply(lambda x: get_city(x).get('country code'))
    df_scraped[f'city_{suffix}_latitude'] = df_scraped[f'city_{suffix}'].apply(lambda x: get_city(x).get('latitude'))
    df_scraped[f'city_{suffix}_longitude'] = df_scraped[f'city_{suffix}'].apply(lambda x: get_city(x).get('longitude'))

df_scraped_usable = df_scraped[df_scraped.city_a_geonameid.notnull()]
df_scraped_usable = df_scraped_usable[df_scraped_usable.city_b_geonameid.notnull()]
df_scraped_usable
```

```python
df_thetis = pd.read_csv(f"thetis_mrv_2019_12_24.pax.with_co2_data.csv")
df_thetis = df_thetis[['imo', 'name']]
df_thetis
```

```python
df_thetis['ship_slug'] = df_thetis.name.apply(slugify)
df_scraped_usable['ship_slug'] = df_scraped_usable.ship.apply(slugify)
df_scraped_usable = df_scraped_usable.set_index('ship_slug').join(df_thetis.set_index('ship_slug'))
```
