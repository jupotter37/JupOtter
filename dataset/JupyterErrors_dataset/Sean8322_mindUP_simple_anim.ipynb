```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

# x = np.arange(0, 10, 0.01)
n=100
# y = [random.uniform(0.75, 0.85) for i in range(10*n)]
y = [random.uniform(0, 1) for i in range(n+50)]
x = list(range(n))
line, = ax.plot(x, y[:n])
mi=0.25
ma=0.4
def specialRand():
    if random.random() < 0.15:
        return random.uniform(mi-0.2, ma+0.2)
    else:
        return random.uniform(mi, ma)
a = np.arange(ma+0.1, mi-0.3, -(ma-mi)/(n+50))
y = [doRand(a[i]-0.05, a[i]+0.05, a[i]) for i in range((n+50))]
y.extend(np.array([doRand(mi-0.1, mi+0.1, a[i]) for i in range((n+50))]))
# y = [random.uniform(mi-0.2*100*n, ma+0.2) for i in range(100*n)]
# y=np.array(y)
x=np.array(x)
# line.set_xdata(np.arange(0, 1, 0.01))
def init():  # only required for blitting to give a clean slate.
    line.set_ydata([np.nan] * len(x))
    return line,


def animate(i):
#     np.delete(y, 0, 0)
    line.set_ydata(y[i:n+i])  # update the data.
    return line,


ani = animation.FuncAnimation(
    fig, animate, init_func=init, interval=100, blit=True, save_count=50)

# To save the animation, use e.g.
#
# ani.save("movie.mp4")
#
# or
#
# writer = animation.FFMpegWriter(
#     fps=15, metadata=dict(artist='Me'), bitrate=1800)
# ani.save("movie.mp4", writer=writer)

ani.save('test.gif')
plt.show()
```
