## Analisis Interquartil

```python

Q1 = df_sess_3.quantile(0.25)
Q1
```

```python

Q3 = df_sess_3.quantile(0.75)
Q3
```

```python

IQR = Q3 - Q1
IQR
```

```python

LOWER_BOUND = Q1 - 1.5*IQR # 1.5 is a standard value
LOWER_BOUND
```

```python

UPPER_BOUND = Q3 + 1.5*IQR
UPPER_BOUND
```

```python

IQR_outliers = df_sess_3[ ((df_sess_3 < LOWER_BOUND) | (df_sess_3 > UPPER_BOUND)).any(axis=1) ]
```

```python

def gender_encoder(gender):
    """
    This function takes a gender as an input
    and returns a codified version of it.
    """
    if gender == "F":
        return 0
    
    elif gender == "M":
        return 1
    
    else:
        return -1
    
```

```python

# df_sess_3["CODE_GENDER"] = gender_encoder(df_sess_3["CODE_GENDER"])
# df_sess_3["CODE_GENDER"] = df_sess_3["CODE_GENDER"].apply(gender_encoder)
df_sess_3["CODE_GENDER"] = df_sess_3["CODE_GENDER"].apply(lambda gender: gender_encoder(gender))
df_sess_3["CODE_GENDER"]
```

# Encode Categorical Features

```python

from sklearn.preprocessing import OneHotEncoder
```

```python

categorical_vars = df_sess_3[["NAME_CONTRACT_TYPE", "FLAG_OWN_CAR", "FLAG_OWN_REALTY", "NAME_TYPE_SUITE", 
                     "NAME_INCOME_TYPE", "NAME_EDUCATION_TYPE", "NAME_EDUCATION_TYPE", "NAME_FAMILY_STATUS", 
                     "NAME_HOUSING_TYPE", "OCCUPATION_TYPE", "ORGANIZATION_TYPE"]]
```

```python

ohenc = OneHotEncoder(handle_unknown='ignore') # Crea el encoder
encoded_labels  = ohenc.fit_transform(categorical_vars) # Nuevas Etiquetas
columns_encoded = ohenc.get_feature_names_out(categorical_vars.columns)   # Diccionario 
categorical_df  = pd.DataFrame(encoded_labels.toarray(), columns=columns_encoded)
```
