## 1) Flocculator (27 points)

A 60 L/s water treatment plant has a vertical flow hydraulic flocculator. The channels are 50 cm wide, 6 m long, and 2 m deep at the end of the flocculator. The head loss through the flocculator is 40 cm. The minimum water temperature is 15°C.
There are 4 channels. The minor loss coefficient for each 180° bend is 2.5.

Answer A through F. 

State any assumptions that you make.

```python
# define your parameters and variables given in the problem statement here
# head loss through the flocculator
headloss_floc = 40 * u.cm

# water depth at the end of flocculator where it flows into the inlet channel of the sedimentation tank
height_floc_end = 2 * u.m

width_floc = 50 * u.cm

length_floc = 6 * u.m

N_channels = 4

flow_floc = 60 * u.L/u.s

# expansion minor loss coefficient for 180 degree bend
K_e = 2.5

# this is the minimum temperature of the raw water
T_water = 15* u.degC
```

### A. (5 points)
What is the hydraulic residence time in the flocculator if you neglect the volume change due to head loss?

```python
volume_floc = width_floc * height_floc_end * length_floc * N_channels
t_floc = (volume_floc / flow_floc).to(u.s)
print('The hydraulic residence time is' , t_floc)
```

### B. (5 points)
What is the average velocity gradient in the flocculator? 

```python
Gave_floc = (np.sqrt(pc.gravity * headloss_floc/(t_floc * pc.viscosity_kinematic(T_water) ))).to(1/u.s)
print('The average velocity gradient in the flocculator is' , Gave_floc)
```

### C. (5 points)
What is the collision potential $\big(G\theta\big)$  for this flocculator?

```python
Gtheta_floc = Gave_floc * t_floc
print('The flocculator collision potential is' , Gtheta_floc)
```

### D. (5 points)
What is the maximum distance between expansions for this flocculator? The maximum ratio of H/S ($\Pi _{H{S_{Max}}}$) is 6

${H_{{e_{Max}}}} = {\bigg[ {\frac{{{K_e}}}{{2\nu {{\bar G}^2}}}{{\left( {\frac{{Q{\Pi _{H{S_{Max}}}}}}{W}} \right)}^3}} \bigg]^{\frac{1}{4}}}$

```python
Pi_HS_max = 6
Lmax_expansions = ((K_e / (2 * pc.viscosity_kinematic(T_water) * Gave_floc**2) * 
                    (flow_floc * Pi_HS_max / width_floc)**3)**(1/4)).to(u.m)
print('The maximum distance between expansions for an efficient design is' , Lmax_expansions)
```

### E. (2 points)
Is it necessary to include additional expansions in this design to maintain high efficiency?

**Your Response Here:**

Additional expansions are not needed in this flocculator since the distance between expansions is greater than the depth of the flocculator.

### F. (5 points)
What is the spacing of the baffles in this flocculator?

$ S = {\left( {\frac{{{K_e}}}{{2{H_e}{G^2}\nu }}} \right)^{\frac{1}{3}}}\frac{Q}{W}$

```python
spacing_baffles = ((K_e / (2 * pc.viscosity_kinematic(T_water) * height_floc_end * Gave_floc**2))**(1/3) * flow_floc / width_floc).to(u.m)
print('The baffle spacing is' , spacing_baffles)
```

## 3) Plate settler design (15 points)
Assume the upflow velocity in AguaClara sedimentation tanks is increased to 3 mm/s,
the capture velocity is increased to 0.2 mm/s, the spacing between plates is maintained at 2.5 cm, and the plates are inclined at 60$^{\circ}$ from the horizontal. You may assume that the vertical velocity entering the plate settlers is 3 mm/s.
Answer questions A and B. 

### A. (10 points)   
How long would the plate settlers be? You may neglect the effect of the thickness of the
plate settlers. **Create a function** for the equation that you use to solve this problem. Include all the necessary inputs in your function definition.

$$L = \frac{{S\left( {\frac{{{V_{Plate \uparrow }}}}{{{V_c}}} - 1} \right)}}{{\sin \alpha \cos \alpha }}$$


```python
V_sed_up = 3 * u.mm/u.s
V_sed_capture = 0.2 * u.mm/u.s
angle_sed_plate = 60 * u.deg
s_sed_plate = 2.5 * u.cm

def L_sed_plate(s_sed_plate, V_sed_up, V_sed_capture, angle_sed_plate):
    return (s_sed_plate * ((V_sed_up/V_sed_capture)-1) / (np.sin(angle_sed_plate) * np.cos(angle_sed_plate))).to(u.m)

sed_plate_length = L_sed_plate(s_sed_plate,V_sed_up,V_sed_capture,angle_sed_plate)

print('The minimum length of the plate settlers is',sed_plate_length)
```

### B. (5 points)   
What is the ** *vertical* ** height of the plate settlers? This will tell us how much of the sedimentation tank depth these plates occupy.

```python
height_sed_plates = L_sed_plate * np.sin(angle_sed_plate) 
print('The vertical height of the plate settlers is',height_sed_plates)
```

## 4) Head loss through a floc blanket (8 points)
### A. (5 points)
What is the head loss through a 1 m deep floc blanket with a clay concentration of 3 g/L and an upflow velocity of  1 mm/s? You may assume the water temperature is 20°C. The density of clay is 2650 $\frac{kg}{m^3}$. This head loss helps ensure a uniform velocity of water leaving the water - floc blanket interface at the top of the floc blanket.

$${h_L}_{_{FB}} = {H_{FB}}\left( {\frac{{{\rho _{Clay}}}}{{{\rho _{{H_2}O}}}} - 1} \right)\frac{{{C_{Clay}}}}{{{\rho _{Clay}}}}$$

```python
T_fb = 20 * u.degC
height_fb = 1 * u.m
density_clay = 2650 * u.kg/u.m**3
Concentration_clay = 3 * u.g/u.L
headloss_fb = (height_fb * (density_clay / pc.density_water(T_fb) - 1) * Concentration_clay / density_clay).to(u.mm)
print('The head loss through the floc blanket is',headloss_fb)
```

## 5) Floc Hopper (10 points)
The AguaClara 1 L/s sedimentation tank has a floc hopper that is a 3" PVC pipe that is angled at 60$^{\circ}$C from the horizontal. The raw water is 1000 NTU and the settled water turbidity is 3 NTU.

A sample from the consolidated sludge in the bottom of the floc hopper was diluted by a factor of 100 and the turbidity of the diluted sample was 150 NTU. (Remember that NTU is an approximate measure of concentration and the unit NTU has been defined for you).

What flow rate must the floc hopper drain valve be set at to maintain a constant level of the consolidated sludge in the floc hopper? 

```python
flow_sed = 1 * u.L/u.s
turbidity_influent = 1000 * u.NTU
turbidity_effluent = 3 * u.NTU
turbidity_sludge = 100 * 150 * u.NTU
sludge_mass_rate = (turbidity_influent - turbidity_effluent) * flow_sed
sludge_flow = (sludge_mass_rate/turbidity_sludge).to(u.mL/u.s)
print('The flow rate of sludge is',sludge_flow)
```
