```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
historical_lcoe = pd.read_csv('/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/data/raw/power_plants_costs/historical_costs/generation.lcoe.20181203_1353.csv')
historical_lcoe
```

```python
historical_lcoe = historical_lcoe[2:]
```

```python
# tech_groups = historical_lcoe.groupby(['Chart Label', 'Year']).mean()
# rowlength = round(tech_groups.ngroups/2+0.5)
# rowlength
# tech_groups
# historical_lcoe[['Chart Label', 'Year','Calculated LCOE']].groupby(['Chart Label', 'Year','Calculated LCOE']).mean()
historical_lcoe[['Year','Calculated LCOE']] = historical_lcoe[['Year','Calculated LCOE']].apply(pd.to_numeric, errors ='coerce', axis=1)
```

```python
# historical_lcoe = historical_lcoe.dropna()
mean_lcoe = historical_lcoe[['Chart Label', 'Year','Calculated LCOE']].groupby(['Chart Label', 'Year']).mean().reset_index()
mean_lcoe['Calculated LCOE'] = mean_lcoe['Calculated LCOE']*1000
rowlength = round(mean_lcoe.groupby('Chart Label').ngroups/2)
mean_lcoe
```

```python
grouped_mean_lcoe = mean_lcoe.groupby("Chart Label")
rowlength = round(grouped_mean_lcoe.ngroups/2)
```

```python
fig, axs = plt.subplots(nrows=rowlength, ncols=2, figsize=(30,30))


targets = zip(grouped_mean_lcoe.groups.keys(), axs.flatten())

for (key, ax) in targets:
    plant_group = grouped_mean_lcoe.get_group(key)#.sort_values('Year', ascending=False)
#     print(plant_group)
    rem_na = plant_group[['Year', 'Calculated LCOE']].dropna()
    ax.plot(rem_na['Year'],rem_na['Calculated LCOE'], marker='o')
    ax.set_title(key)
ax.legend()
plt.show()
```

## Simplified Type

```python
list(mean_lcoe['Chart Label'].unique())
```

```python
simplified_type = mean_lcoe.copy()
simplified_type['Simplified_type'] = simplified_type['Chart Label'].map(lambda x: "Coal" if "Scrubbed" in x 
                                                                        else "Coal" if "Unscrubbed" in x
                                                                        else "Biomass" if "Biopower" in x
                                                                        else "CCGT" if "Combined Cycle" in x
                                                                        else "OCGT" if "Combustion Turbine" in x
                                                                        else "Hydro" if "Hydroelectric" in x
                                                                        else "Nuclear" if "Nuclear" in x
                                                                        else "Offshore" if "Offshore" in x
                                                                        else "Onshore" if "Onshore" in x
                                                                        else "PV" if "Photovoltaic" in x
                                                                        else float('NaN'))

simplified_type = simplified_type.dropna().set_index("Simplified_type").drop("Chart Label", axis=1)
simplified_type
```
