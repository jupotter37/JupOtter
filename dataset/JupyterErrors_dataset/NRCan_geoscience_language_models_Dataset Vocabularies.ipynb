## Get the list of all generated datasets...

The base path to the datasets will need to be updated to reflect your system. 

```python
import pathlib 
```

```python
datasets = []

file_list = []
for folder in ['v4', 'v4_A_B', 'v4_A_B_D']:
    for file in pathlib.Path('/nrcan_p2/data/03_primary/', folder).iterdir():
        if file.is_file():
            if file.suffix == ".txt" and 'partial' not in str(file):
                if 'dA_full' in str(file):
                    file_list.append(file)
 

for file in file_list:
    print(file)
#print(file_list)
                
```

## Read the associated vocab files for each of the datasets...

```python
import pandas as pd

def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1

def count_words_in_file(fname):
    with open(fname) as f:
        nwords_total = 0
        for l in f:
            nwords = l.strip()
            if len(nwords) == 0:
                continue
            nwords = nwords.split()
            nwords_total += len(nwords)
    return nwords_total

vocab_len = {}
for file in file_list:
    print(file)
    vocab_file_name = file.parent / ("vocab_" + file.stem + "_None.csv")
    vocab_file_name2 = file.parent / ("vocab_" + file.stem + "_rm_punct_lower.csv" )       
            
    vocab_len[file] = {}
    vocab_len[file]['n_words'] = count_words_in_file(file)
    if vocab_file_name.exists():
        vocab_len[file]['None'] = file_len(vocab_file_name)
        
    if vocab_file_name2.exists():
        vocab_len[file]['rm_punct_lower'] = file_len(vocab_file_name2) 
        vocab_len[file]['n_sent'] = file_len(file)
        
vocab_len
```

### Display the vocab stats in a formatted dataframe...

```python
vocab_df = pd.DataFrame.from_dict(vocab_len, orient='index').reset_index()
vocab_df['dataset'] = vocab_df['index'].apply(lambda x: 'A' if 'dA_full_v1' in str(x) else 'A+B' if 'dA_full_dB_v1' in str(x) else 'A+B+D')
vocab_df['pipeline'] = vocab_df['index'].apply(lambda x: 'v1' if 'SIMPLE' in str(x) else '80' if '80' in str(x) else '90' if '90' in str(x) else 'PLUS')
vocab_df['pipeline_model'] = vocab_df['index'].apply(lambda x: 'GLOVE' if 'GLOVE' in str(x) else 'BERT')
#vocab_df
x = vocab_df[['pipeline_model','dataset', 'pipeline', 'rm_punct_lower', 'None', 'n_words', 'n_sent']].sort_values(['pipeline_model', 'dataset', 'pipeline']).set_index(['pipeline_model', 'dataset', 'pipeline'])
x.to_csv('vocab_dataset_stats.csv')
x
```
