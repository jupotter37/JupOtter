```python
def esp_clients_per_week(mean=20.433962264150942, std=3.5432472792051746):
        """This generates the number of new clients in ESP for a given week.
        The default parameters are taken from the years 2013-2016."""
        oneweek_esp_clients = round(stats.norm.rvs(mean,std))
        if oneweek_esp_clients <0:
            oneweek_esp_clients = 0
        oneweek_esp_clients = oneweek_esp_clients 
        return oneweek_esp_clients
```

```python
sns.kdeplot(np.array([esp_clients_per_week() for i in range(500)]))
plt.title('Clients per week')
```
