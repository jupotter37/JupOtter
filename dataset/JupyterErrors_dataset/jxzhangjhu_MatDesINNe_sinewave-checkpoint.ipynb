```python
from numpy import sin, cos
import matplotlib.pyplot as plt
import numpy as np
from smt.sampling_methods import LHS

f = 3
# plot the contour of sinewave function
delta = 0.01
x = np.arange(-1.0, 1.0, delta)
y = np.arange(-1.0, 1.0, delta)
X, Y = np.meshgrid(x, y)
Z = sin(f*np.pi*X)  +  cos(f*np.pi*Y)

fig, ax = plt.subplots(figsize=(10,10))
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')
plt.savefig('true.png', dpi=None)

## generate data using latin hypercube sampling

xlimits = np.array([[0.0, 4.0], [0.0, 3.0]])
sampling = LHS(xlimits=xlimits)
num =100000
x = sampling(num)

fig, ax = plt.subplots(figsize=(10,10))
plt.plot(x[:, 0], x[:, 1], "o")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


data_x = x 
data_y = sin(f*np.pi*x[:,0]) + cos(f*np.pi*x[:,1])
np.savetxt('data_x.csv', data_x, delimiter=',')
np.savetxt('data_y.csv', data_y, delimiter=',')
```
