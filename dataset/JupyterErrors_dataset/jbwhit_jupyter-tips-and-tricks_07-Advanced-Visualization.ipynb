```python
x = np.linspace(0, 2, 50)
y = x**2
```

```python
fig = plt.figure()
scatter = plt.scatter(x, y)
plt.show()
```

```python
fig.animation_duration = 500
scatter.y = x**.5
```

```python
scatter.selected_style = {'stroke':'red', 'fill': 'orange'}
plt.brush_selector();
```

```python
scatter.selected
```

```python
scatter.selected = [1,2,10,42,45]
```

## ipyvolume

```python
import ipyvolume as ipv
```

```python
N = 1000
x, y, z = np.random.random((3, N))
```

```python
fig = ipv.figure()
scatter = ipv.scatter(x, y, z, marker='box')
ipv.show()
```

```python
scatter.x = x + np.random.rand() 
```

```python
scatter.color = "green"
scatter.size = 5
```

```python
scatter.color = np.random.random((N,3))
```

```python
scatter.size = 2
```

```python
ex = ipv.datasets.animated_stream.fetch().data
```

```python
ipv.figure()
ipv.style.use('dark')
quiver = ipv.quiver(*ipv.datasets.animated_stream.fetch().data[:,::,::4], size=5)
ipv.animation_control(quiver, interval=200)
ipv.show()
ipv.style.use('light')
```

```python
quiver.geo = "cat"
```
