```python
import Data.StockDataManager as sm
import Data.TimeSeries as ts

settings = sm.Settings()
tmgr = sm.TaskManager(settings)

data = ts.TimeSeries(settings).get_stock_series(['000001'], start='2012-01-01', fields=['close'])
price = data['close']
```

```python
import matplotlib 
%matplotlib inline
```

```python
from pyhht.visualization import plot_imfs
from pyhht.emd import EMD

import numpy as np

signal = np.array(price)
imfs = EMD(signal).decompose()
plot_imfs(signal, np.linspace(0, 1, len(signal)), imfs)
```
