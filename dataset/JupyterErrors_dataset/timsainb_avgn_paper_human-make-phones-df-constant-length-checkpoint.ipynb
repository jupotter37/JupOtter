### create dataset

```python
hparams = HParams(
    num_mel_bins = 128,
    mel_lower_edge_hertz=100,
    mel_upper_edge_hertz=7998,
    butter_lowcut = 100,
    butter_highcut = 7998,
    ref_level_db = 20,
    min_level_db = -50,
    mask_spec = True,
    win_length_ms = 10.0,
    hop_length_ms = 0.25,
    n_jobs = -1,
    verbosity=1,
    nex = -1
)
```

```python
# create a dataset object
dataset = DataSet(DATASET_ID, hparams = hparams)
```

```python
n_jobs = -1
verbosity=1
```

```python
with Parallel(n_jobs=n_jobs, verbose=verbosity) as parallel:
    syllable_dfs = parallel(
        delayed(create_label_df)(
            dataset.data_files[key].data,
            hparams=dataset.hparams,
            labels_to_retain=[
                "labels",
                "word",
                "utterance_number",
                "position_in_utterance",
                "position_in_word",
                "word_number",
            ],
            unit="phones",
            dict_features_to_retain=["age", "sex"],
            key=key,
        )
        for key in tqdm(dataset.data_files.keys())
    )
syllable_df = pd.concat(syllable_dfs)
print(len(syllable_df))
syllable_df[:3]
```

```python
syllable_df = syllable_df[syllable_df.end_time.values - syllable_df.start_time.values < 3]
syllable_df = syllable_df[syllable_df.end_time.values - syllable_df.start_time.values > 0.005]
```

### get audio for dataset

```python
with Parallel(n_jobs=n_jobs, verbose=verbosity) as parallel:
    syllable_dfs = parallel(
        delayed(get_row_audio)(
            syllable_df[syllable_df.key == key], 
            dataset.data_files[key].data['wav_loc'], 
            dataset.hparams
        )
        for key in tqdm(syllable_df.key.unique())
    )
syllable_df = pd.concat(syllable_dfs)
print(len(syllable_df))
syllable_df[:3]
```

```python
sylls = syllable_df.audio.values
nrows = 5
ncols = 10
zoom = 2
fig, axs = plt.subplots(ncols=ncols, nrows = nrows,figsize = (ncols*zoom, nrows+zoom/1.5))
for i, syll in tqdm(enumerate(sylls), total = nrows*ncols):
    ax = axs.flatten()[i]
    ax.plot(syll)
    if i == nrows*ncols -1:
        break
```

```python
# remove short syllables
syll_lens = np.array([len(i) for i in tqdm(syllable_df.audio.values)])
syllable_df = syllable_df[syll_lens > 100]
```

```python
# normalize audio
syllable_df['audio'] = [(i/np.max(np.abs(i))).astype('float32') for i in tqdm(syllable_df.audio.values)]
```

```python
syllable_df['audio'].values[0]
```

### Create spectrograms

```python
with Parallel(n_jobs=n_jobs, verbose=verbosity) as parallel:
    # create spectrograms
    syllables_spec = parallel(
        delayed(make_spec)(
            row.audio,
            row.rate,
            hparams=dataset.hparams,
            mel_matrix=dataset.mel_matrix,
            use_mel=True,
            use_tensorflow=False,
            norm_uint8 = True,  
        )
        for idx, row in tqdm(
            syllable_df.iterrows(),
            total=len(syllable_df),
            desc="getting syllable spectrograms",
            leave=False,
        )
    )
```

```python
syll_lens = [np.shape(i)[1] for i in syllables_spec]

np.max(syll_lens)

plt.hist(syll_lens, bins=500);
```

### Resize spectrogram

```python
from PIL import Image

def resize_spec(spec, size=128):
    resize_shape = [size, np.shape(spec)[0]]
    resize_spec = np.array(Image.fromarray(spec).resize(resize_shape, Image.ANTIALIAS))
    return resize_spec
```

```python
with Parallel(n_jobs=n_jobs, verbose=verbosity) as parallel:
    syllables_spec = parallel(
        delayed(resize_spec)(spec, size=128)
        for spec in tqdm(syllables_spec, desc="scaling spectrograms", leave=False)
    )
```

```python
def norm(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x))

def remove_med(x):
    x[x<=np.median(x)] = 0
    return x

syllables_spec = [(norm(i)*255).astype('uint8') for i in tqdm(syllables_spec)]
```
