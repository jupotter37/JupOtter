```python
import numpy as np
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
#sns.set_style("darkgrid")
plt.rcParams["axes.grid"] = True


x=[ 20, 0.51, 0.1, 0.45, ] # Num parameters
y=[ 0.86, 0.85 , 0.76, 0.72] # minADE (k=6)

z=["WIMP" ,"CAT-Pred" , "Ours-Social","Ours-Social+Map", "LaneGCN,w/o map", "LaneGCN", "HOME", "GOHOME"]

palette=['green', "violet", "orange", "red", "blue"]
#palette=[ "blue", "blue", "blue", "blue", "blue", "blue", "red"]

non_arrow_indices = {0,1,2,3,4,5,6} #{2,5}
displacements = [(0,0), (0,0), (0,0)]

# create pandas dataframe
data = pd.DataFrame(
    {'N. params (M)': x,
     'minADE': y,
     'Method': z,
    })

display(data)


plt.figure(figsize=(12,6))

markers = ["p", "P", "o", "X", "D"]
print (len(x), len(markers))

for idx, (_x,_y) in enumerate(zip(x,y)):
    plt.scatter(_x,_y, color=palette[idx], marker=markers[idx], s=400)

plt.xlabel("N. params (M)", fontsize=30)
plt.ylabel("minADE (m)", fontsize=30)
plt.ylim(0, 1)
#plt.yticks(np.arange(0.9, 1,0.05), fontsize=40)
plt.xlim(0.05,21)
#plt.xticks(np.arange(0,1000,100), fontsize=40)


font = {'family' : 'normal',
        #'weight' : 'bold',
        'size'   : 20}

matplotlib.rc('font', **font)

# adnotate the samples in the plot 
for i, method in enumerate(z):
    dx, dy = displacements[i]

    if "ours" in method:
        plt.annotate(method,(x[i] + dx, y[i] + dy), weight='bold')
    else:
        plt.annotate(method,(x[i] + dx, y[i] + dy)) # weight='bold'
    

plt.tight_layout()
plt.savefig("teaserv2.png") 
plt.savefig("teaserv2.pdf")
plt.show()
```
