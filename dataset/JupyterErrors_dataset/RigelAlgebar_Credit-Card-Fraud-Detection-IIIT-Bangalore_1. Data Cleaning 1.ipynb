```python

def percentage_nan(data: pd.DataFrame):
    """
    This function calculates the percentage 
    of nan values within all of the columns 
    of a dataframe.
    """
    
    columns = list(data.columns)
    percentages = {col:0 for col in columns} # Dictionary Comprehension
    
    for col in columns:
        percentages[col] = 100*data[col].isna().sum()/len(data.index)
        
    
    return percentages
    
```

```python

missing_vals_percentages = percentage_nan(df_applications)
missing_vals_percentages
```

```python

plt.figure(figsize=(20,12))
plt.bar(missing_vals_percentages.keys(), missing_vals_percentages.values())



plt.xticks(rotation=90)
plt.yticks(fontsize=15)

plt.xlabel("Variables", fontsize=25)
plt.ylabel("Percentage of Missing Values", fontsize=25)
plt.show()
```

```python

for key,val in missing_vals_percentages.items():
    print(key, val)
```

```python

cols_missing_vals = {key: val for key,val in missing_vals_percentages.items() if val>0}
cols_missing_vals
```

```python

plt.figure(figsize=(20,12))
plt.bar(cols_missing_vals.keys(), cols_missing_vals.values())



plt.xticks(rotation=90)
plt.yticks(fontsize=15)

plt.xlabel("Variables", fontsize=25)
plt.ylabel("Percentage of Missing Values", fontsize=25)
plt.show()
```

```python

THRESHOLD = 40

cols_to_remove = [key for key, val in cols_missing_vals.items() if val>=THRESHOLD] # List Comprehension
cols_to_remove
```

```python

df_apps_clean = df_applications.drop(columns=cols_to_remove)
```
