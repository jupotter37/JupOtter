````python
x_0, y_0 = var('x_0 y_0', domain="real")
````

````python
P = vector([x_0, y_0])
````

````python
k = var('k', domain="real")
lamb = var('lamb', domain="real")
#assume(lamb > 0)
X_lambda = vector([1., lamb])
````

````python
Q_lambda = P + k* X_lambda
print(Q_lambda)
````

````python
EllipsePolynomial = (x^2 / a^2 + y^2 / b^2 - 1)
print(EllipsePolynomial)
print(type(EllipsePolynomial))
````

````python
EllipsePolynomial.subs(x == Q_lambda[0]).subs(y == Q_lambda[1])
````

````python
k_solutions = solve(EllipsePolynomial.subs(x == Q_lambda[0]).subs(y == Q_lambda[1]), k)
````

````python
k_solutions[0].full_simplify()
````

````python
Q_lambda[0].subs(k == k_solutions[0].rhs())
````

````python
Q_lambda[1].subs(k == k_solutions[0].rhs())
````
