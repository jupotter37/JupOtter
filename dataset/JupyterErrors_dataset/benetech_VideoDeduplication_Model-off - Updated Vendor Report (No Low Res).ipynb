## Evaluation by cluster predominance / purity

```python
from sklearn.cluster import DBSCAN
```

Scenario 1: Clusters are assigned according to most predominant label

```python
augmentation_manifesto.head()
```

```python
distances = np.linspace(0.0001,0.5)
results = []
n_clusters = []
for d in distances:
    
    clusters = DBSCAN(metric='cosine',eps=d,min_samples=1).fit_predict(video_space)
    n_clusters.append(len(np.unique(clusters)))
    model_results = pd.DataFrame({'filename':labels,'cluster':clusters})
    filename_original = a_o = dict(zip(augmentation_manifesto['filename_x'].apply(lambda x:x.split('.')[0]),augmentation_manifesto['original_filename']))
    filename_original = a_o = dict(zip(augmentation_manifesto['filename_x'].apply(lambda x:x.split('.')[0]),augmentation_manifesto['original_filename']))
    model_results['original_basename_q'] = model_results['filename'].apply(lambda x:matching(x,filename_original))
    model_results['original_basename_q'] = model_results['original_basename_q'].apply(lambda x:x.split('.')[0])
    model_results ['original_basename_m'] = 'NA'
    for c in model_results.cluster.unique():

        msk = model_results.cluster == c
        prevailing = model_results.loc[msk,:].original_basename_q.value_counts().index.values[0]
        model_results.loc[msk,'original_basename_m'] = prevailing


    per_a = model_results['original_basename_q'].value_counts()
    msk = per_a[per_a > 1].index.values
    restricted = model_results.loc[model_results['original_basename_q'].isin(msk),:]   
    result = sum(restricted['original_basename_m'] == restricted['original_basename_q']) / restricted.shape[0]   
    results.append(result)
        
```

```python
plt.plot(distances,results)
plt.xlabel('Distance')
plt.ylabel('Accuracy')
plt.title('Lax Accuracy vs Distance')
```

This analysis presents a very serious flaw, since the predominance criteria (or purity) assigns labels to clusters of length 1, this means if there is only one file for a given cluster number, it will be assigned the correct label and be counted towards the "lax" accuracy criteria.

```python
maximum_unique_files = model_results['original_basename_q'].nunique()
plt.plot(distances,n_clusters)
plt.plot(distances,[len(video_space) for x in distances])
plt.plot(distances,[maximum_unique_files for x in distances])
plt.xlabel('Distance')
plt.ylabel('Number of Clusters')
plt.title('Clusters vs Distance')
```

Orange line represents the maximum number of files / Green Line represents the maximum number of unique files 

note: I refer to it as the maximum number of files because the original seed files were not labeled, so there might be duplicated files within the original dataset.

## What if we restricted the problem only assigned labels for clusters with length > 1

```python
distances = np.linspace(0.0001,0.5)
results = []
for d in distances:
    
    clusters = DBSCAN(metric='cosine',eps=d,min_samples=1).fit_predict(video_space)
    model_results = pd.DataFrame({'filename':labels,'cluster':clusters})
    filename_original = a_o = dict(zip(augmentation_manifesto['filename_x'].apply(lambda x:x.split('.')[0]),augmentation_manifesto['original_filename']))
    filename_original = a_o = dict(zip(augmentation_manifesto['filename_x'].apply(lambda x:x.split('.')[0]),augmentation_manifesto['original_filename']))
    model_results['original_basename_q'] = model_results['filename'].apply(lambda x:matching(x,filename_original))
    model_results['original_basename_q'] = model_results['original_basename_q'].apply(lambda x:x.split('.')[0])
    model_results ['original_basename_m'] = 'NA'
    for c in model_results.cluster.unique():
        msk = model_results.cluster == c
        prevailing = model_results.loc[msk,:].original_basename_q.value_counts().index.values[0]
        top = model_results.loc[msk,:].original_basename_q.value_counts().values[0]
        if top != 1:
            model_results.loc[msk,'original_basename_m'] = prevailing
        else:
            model_results.loc[msk,'original_basename_m'] = 'unique'
            
    per_a = model_results['original_basename_q'].value_counts()
    msk = per_a[per_a > 1].index.values
    msk_unique = per_a[per_a == 1].index.values
    
    model_results['original_basename_q_modified'] = model_results['original_basename_q']
    model_results.loc[model_results['original_basename_q'].isin(msk_unique),'original_basename_q_modified'] = 'unique'
    
    result = sum(model_results['original_basename_m'] == model_results['original_basename_q_modified']) / model_results.shape[0]
    
#     model_results.loc[msk_unique]
    
#     restricted = model_results.loc[model_results['original_basename_q'].isin(msk),:]   
#     result = sum(restricted['original_basename_m'] == restricted['original_basename_q']) / restricted.shape[0]   
    results.append(result)
    
print(results)
```

## Query Based

```python
from sklearn.neighbors import NearestNeighbors,KDTree,BallTree,LSHForest,NearestCentroid
from pyvis.network import Network
```

```python
def filter_results(distances,indexes,thr):
    results = []
    results_distances = []
    msk = distances < thr
    for i,r in enumerate(msk):
        results.append(indexes[i,r])
        results_distances.append(distances[i,r])
    return results,results_distances



def evaluate_match_based(estimator,plot=True,thr=0.4):
    
    estimator.fit(video_space)
    distances,indexes = estimator.kneighbors(video_space,n_neighbors=20)
    if plot == True:
        plt.hist(distances[:,1::].ravel(),bins=100)
        
    
    results,results_distances = filter_results(distances,indexes,thr)
    
    net = Network(notebook=True,height='1000px',width='1000px')
    net.barnes_hut()

    for i,l in enumerate(labels):
        net.add_node(str(i),label=str(l))

    for i,d in enumerate(results):
        for j,m in enumerate(d):
            if j != 0:
                net.add_edge(str(i),str(m),weight=str(results_distances[i][j]))

    net.show_buttons(filter_=['physics'])
    net.show('mygraph.html')
    return net
    
    
```
