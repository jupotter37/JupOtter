# Morse Theory

Consider   

$ 
f(x) = f(a) + (x^1)^2 + (x^2)^2 + \dots + (x^{\mu})^2  - (x^{\mu +1})^2 + (x^{\mu+2})^2 + \dots + (x^n)^2 
$  
   
for $x \in \mathbb{R}^n$   

So $\mu$ is the so-called **Morse index**.




## $\mathbb{R}^2$

```python
x_2,y_2 = var('x_2,y_2')  # coordinates 1,2 for x^1, x^2 or x, y for R2
```

```python
f0_2(x_2,y_2) = -x_2**2 - y_2**2  # Morse index of 0, for R2
f1_2(x_2,y_2) = x_2**2 - y_2**2  # Morse index of 1, for R2
f2_2(x_2,y_2) = x_2**2 + y_2**2  # Morse index of 2, for R2
```

`implicit_plot` of Sage Math plots the curve $f(x,y)=0$

## $\mathbb{R}^3$

```python
x,y,z = var('x,y,z')  # coordinates 1,2,3 for x^1, x^2, x^3 or x, y, z for R3
```

## Using `sagemanifolds` for Morse Theory

```python
%%latex 
```

```python
M3 = Manifold(3,'M3',latex_name=r'\mathcal{M}^3', start_index=1)
```

```python
X.<x,y,z> = M3.chart() # global chart
```

```python
#f3 = M3.scalar_field( function('f3',  *[X[i] for i in M3.irange()]),  name='f3', latex_name=r'f^{(3)}') # deprecated form for function
f3 = M3.scalar_field( function('f3')(*[X[i] for i in M3.irange()]),  name='f3', latex_name=r'f^{(3)}')
```

```python
f3.differential().display()
```

```python
f3.differential().exterior_derivative().display()
```

So I'll try another way to construct the **Hessian**

```python
M3.scalar_field_algebra()
```

Construct $f \in C^{\infty}(M)$ such that $\text{dim}{M} =3$ (just in this case), and I'll call $f$ to be `f3` in the Python code (i.e. notation).

So, cf. [
Algebra of Differentiable Scalar Fields](http://sagemanifolds.obspm.fr/doc/reference/manifolds/sage/manifolds/differentiable/scalarfield_algebra.html?highlight=scalar_field_algebra)

```python
f3 = M3.scalar_field_algebra()( function('f3')(*[X[i] for i in M3.irange()]) )
```

```python
f3.differential().display()
```

Note that we can "call" for the components by the usual Python-like indexing (i.e. with [1], [2], etc.)

```python
f3.differential()[1]
```

```python
M3.scalar_field_algebra()( f3.differential()[1] ).differential().display()
```

```python
M3.scalar_field_algebra()( f3.differential()[1] ).differential()[:]
```

Construct $\mathcal{X}(M)$ with $\text{dim}(M)=3$ (in this case).  In the Python code below, it'll be denoted `XM3`

```python
XM3 = M3.vector_field_module()
```

```python
Hessian3 = XM3.tensor_from_comp((1,1), 
                                [
        M3.scalar_field_algebra()(f3.differential()[i]).differential()[:] for i in M3.irange()
    ], 
                               name='Hessian3',latex_name='D(D(f^{(3)}))' )
```

```python
ddf11 = M3.scalar_field_algebra()( f3.differential()[1]).differential()[:][0];
print(ddf11)
XM3.tensor_from_comp( (1,1), [[ddf11,0,0],[0,0,0],[0,0,0]])
```

So if you get `AttributeError: 'list' object has no attribute '_ring'`, then that means you have to use Components, in class sage.tensor.modules.comp.Components

```python
from sage.tensor.modules.comp import Components
```

```python
comp3 = Components( M3.scalar_field_algebra(), X.frame(), 
                   nb_indices=2, # nb_indices - number of indices labeling componenets
                   start_index=1,output_formatter=XM3._output_formatter)
```

```python
for i in M3.irange():
    for j in M3.irange():
        comp3[i,j] = M3.scalar_field_algebra()( f3.differential()[i]).differential()[:][j-1]
# sanity check
#for i in M3.irange():
#    print i, M3.scalar_field_algebra()( f3.differential()[i]).differential()[:][1]
# Sanity check: this works comp3[1,1,1] = ddf11
```

```python
# sanity check this works  [ M3.scalar_field_algebra()(f3.differential()[i]).differential()[:] for i in M3.irange() ]
```

```python
# sanity check
# help(comp3.display)
comp3.display('H')
```

```python
Hessian3 = XM3.tensor_from_comp((1,1), 
                                comp3, 
                               name='Hessian3',latex_name='D(D(f^{(3)}))' )
```
