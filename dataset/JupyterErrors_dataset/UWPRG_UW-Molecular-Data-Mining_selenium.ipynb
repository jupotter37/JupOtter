```python
# download the chromedriver for the version you are using.
driver = webdriver.Chrome('/Users/wab665/Downloads/chromedriver')

#open google scholar, ylo=2015
driver.get("https://scholar.google.com/scholar?as_ylo=2015&q=flame+retardant&hl=en&as_sdt=0,48")

elems = driver.find_elements_by_xpath("//a[@href]")
hits = []
our_html_links = []
for index, elem in enumerate(elems):
    link = elem.get_attribute("href")
    if 'scholar' not in link and\
        'void' not in link:
        hits.append(index)
        if len(hits) > 1:             #after filtering for 'scholar' and 'void' we get the two main links of each article,
            if index - hits[-2] > 5:  #the page links and the full article links, which appear w/in 5 indices (usually)
                if 'pdf' not in link: #its the very next link but sometimes not
                    our_html_links.append(index)
for index in our_html_links:
    elem = elems[index]
    elem.click()
    driver.back()
    elems = driver.find_elements_by_xpath("//a[@href]")
# elem = driver.find_element_by_name("records_chunks")
# elem.clear()
# elem.send_keys("pycon")
# elem.send_keys(Keys.RETURN)
```
