### 0s and null values

```python
# find all 0s in total columns
cols = ACS.columns.to_list()
for col in cols:
    if  col.find('tot') != -1:
        # show miss
        print(col, ACS[str(col)].isin([0]).sum())
```

Why do *educ_tot_pop* and *total_pop_16_plus*  have fewer 0s than *tot_population*?
Because *tot_population* is short for "total population whose poverty status could be determined."
Check to confirm:
- other 'total' variables have 0s iff *total_pop_16_plus* has a zero

```python
for col in cols:
    if  col.find('tot') != -1:
        print(col, ACS.loc[(ACS[str(col)] != 0) & (ACS.total_pop_16_plus == 0), :].shape)
```
