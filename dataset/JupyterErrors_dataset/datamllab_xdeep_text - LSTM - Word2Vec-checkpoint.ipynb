```python
import os
import numpy as np
from sklearn import model_selection
from sklearn.pipeline import make_pipeline
from keras.models import load_model
from keras.preprocessing.text import one_hot


def load_polarity(path='/Users/zhangzijian/Downloads/rt-polaritydata/rt-polaritydata'):
    data = []
    labels = []
    f_names = ['rt-polarity.neg', 'rt-polarity.pos']
    for (l, f) in enumerate(f_names):
        for line in open(os.path.join(path, f), 'rb'):
            try:
                line = line.decode('latin-1')
            except:
                print(line)
            data.append(line.strip())
            labels.append(l)
    return data, np.array(labels)


vocab_size = 15700
raw_data, labels = load_polarity()
data = [one_hot(d, vocab_size) for d in raw_data]
train, test, train_labels, test_labels = model_selection.train_test_split(raw_data, labels, test_size=.2, random_state=42)
train, val, train_labels, val_labels = model_selection.train_test_split(train, train_labels, test_size=.1, random_state=42)

model = load_model("lstm.h5")
```

```python
import xdeep.local.perturbation.xdeep_text as xdeep_text
%load_ext autoreload
%autoreload 2

def predict(x):
    if len(x) == 1:
        x = one_hot(x, vocab_size)
    else:
        x = [one_hot(d, vocab_size) for d in x]
    model.predict(x)
    
explainer = xdeep_text.TextExplainer(predict, ['negative', 'positive'])
```

```python
def show_explanation(idx, dataset, labels):
    text = dataset[idx]
    print("Ground Truth:", labels[idx])

    explainer.explain('lime', text)
    explainer.show_explanation('lime')
    explainer.explain('cle', text, spans=(2,))
    explainer.show_explanation('cle')
```
