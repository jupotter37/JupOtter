```python
%%script glpsol -m /dev/stdin 

# Read CSV data file

set S dimen 2;
param d{S};
param c{S};

table data IN "CSV" "data.csv":
    S <- [FROM,TO], d~DISTANCE, c~COST;
    
# Construct sets

set FROM := setof{(a,b) in S} a;
set TO := setof{(a,b) in S} b;
set CITIES := FROM union TO;
    
# Minimum Distance
var x{S} binary;
minimize cost : sum{(a,b) in S} x[a,b]*c[a,b];

s.t. A{t in TO}: sum{f in FROM} x[f,t] = 1;



solve;

for {(a,b) in S} : printf "%12s %12s %d\n",a,b,x[a,b];

for {a in CITIES} : printf "%s\n",a;
    
end;
```

```python
from cStringIO import StringIO
import PIL
import urllib
from IPython.display import Image

from geopy import geocoders
g = geocoders.GoogleV3()

place,(lat,lng) = g.geocode("52044 Carding Mill Ct., Granger, IN 46530")

zoom = 10
hsize = 640
vsize = 320
scale = 2

url = "http://maps.googleapis.com/maps/api/staticmap" + \
    "?center={0},{1}".format(lat,lng) + \
    "&size={0}x{1}".format(hsize,vsize) + \
    "&scale={0}".format(scale) + \
    "&zoom={0}".format(zoom) + \
    "&maptype=roadmap&sensor=false"
        
buffer = StringIO(urllib.urlopen(url).read())
im = PIL.Image.open(buffer).convert()

figure(figsize=(15,15))
fig = imshow(im)
axis('off')
scatter([hsize*scale/2],[vsize*scale/2])
url
```
