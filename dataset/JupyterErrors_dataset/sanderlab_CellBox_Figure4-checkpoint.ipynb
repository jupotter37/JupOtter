```python
wn_full = pd.read_csv('whitenoise_correlation_full.csv',index_col=0)
```

```python
wn_full['train_mse']=np.nan
for i in range(len(wnoises)):
    noise = wnoises[i]
    indxs = np.where(wn_full['noise']==noise)[0]
    path = glob.glob('/n/data1/hms/cellbio/sander/bo/CellBox/CellBox/results/WhiteNoise_Jun17_'+str(i)+'_*')
    folders = os.listdir(path[0])
    #folders.sort()
    if len(indxs) != len(folders):
        print('Error!')
        break
    else:
        for j in range(len(indxs)):
            file = glob.glob(path[0]+'/'+folders[j]+'/record_eval.csv')
            if len(file) == 1:
                #trainmse = float(file[0][file[0].find('loss')+5:file[0].find('.csv')])
                record = pd.read_csv(file[0],index_col=False)
                try:
                    trainmse = np.mean([float(record['train_mse'].iloc[-x]) for x in range(3,15)])
                    wn_full['train_mse'].iloc[indxs[j]]=trainmse
                except:
                    continue
```

```python
wn_full['pass']=[wn_full['train_mse'].iloc[x]<0.05 and wn_full['dxdt'].iloc[x] < 0.001 for x in range(len(wn_full))]
```
