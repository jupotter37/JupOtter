```
import os
import cv2
```

```
ls
```

```
images=[]
path=os.listdir('images')
ii=0
for im in path:
  print(ii)
  image=cv2.imread('images/'+im)
  image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  image=cv2.resize(image,(224,224))
  image=image/255
  images.append(image)
  ii=ii+1
```

```
images
```

```
len(images)
```

```
from keras.applications import VGG16
```

```
vgg16=VGG16()
```

```
vgg16.summary()
```

```
lastlayer=vgg16.get_layer('fc2')
```

```
from keras.models import Model
```

```
model=Model(input=vgg16.input,output=lastlayer.output)
```

```
model.summary()
```

```
import pickle
```

```
pickle.dump(model,open('model.pkl','wb+'))
```

```
import numpy as np
```

```
images=np.array(images)
```

```
features=model.predict(images)
```

```
features
```

```
pickle.dump(features,open('features.pkl','wb+'))
```

```
features.shape
```

```
from scipy.spatial.distance import cosine
```

```
ls
```

```
image=cv2.imread('/content/_97265525_20616979_10155064962591026_6934617954464082233_o.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image=cv2.resize(image,(224,224))
image=image/255
  
```

```
query_feature=model.predict(np.array([image]))
```

```
query_feature
```

```
cosine_distance=[]
ii=0
for i in features:
  dis=cosine(query_feature,i)
  cosine_distance.append([dis,ii])
  ii=ii+1
```

```
cosine_distance
```

```
cosine_similarity=sorted(cosine_distance)
```
