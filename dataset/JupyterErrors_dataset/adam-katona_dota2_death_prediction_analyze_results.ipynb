```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve

import pandas as pd

import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
epoch = np.load('epoch_per_sec_predictions.npy')
```

```python
all_pred = []
all_labels = []
label_treshold = 5
for i,vec in enumerate(epoch):
    all_pred.extend(vec)
    if i >= label_treshold:
        all_labels.extend([0]*len(vec))
    else:
        all_labels.extend([1]*len(vec))
```

```python
plt.rcParams['figure.figsize'] = [9, 9]
```

```python
fpr, tpr, thresholds = roc_curve(all_labels, all_pred)

plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr)
```

```python
precision, recall, thresholds = precision_recall_curve(all_labels, all_pred)
plt.plot(recall, precision)
```
