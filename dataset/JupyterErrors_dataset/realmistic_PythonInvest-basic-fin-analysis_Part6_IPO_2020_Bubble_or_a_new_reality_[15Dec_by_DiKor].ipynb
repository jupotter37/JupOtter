```
# Function for cleaning the data to numbers - similar to what we used in the prev. tutorial

# https://stackoverflow.com/questions/7312020/why-wont-re-groups-give-me-anything-for-my-one-correctly-matched-group
import re

POWERS = {'T': 10 ** 12,'B': 10 ** 9, 'M': 10 ** 6, '%': 0.01, '1':1}

"""Read a string (with M/B/T/% values)
   Return a correct numeric value
"""
def convert_str_to_num(num_str):
   match = re.search(r"([0-9\.-]+)(M|B|T|%)?", num_str)
   if match is None:
     print(num_str)
     return None
   else:   
     quantity = match.group(1)
     if match.group(2) is None:
       magnitude = '1' # no modificator in the end -> don't multiply on anything
     else:  
       magnitude = match.group(2)
    #  print(quantity, magnitude)
     return float(quantity) * POWERS[magnitude]
```

```
# Testing the function : return all numeric values, except cases when can't convert (not a number)
print(convert_str_to_num("$2362.5"))
print(convert_str_to_num("+0.64%"))
print(convert_str_to_num("-0.0600"))
print(convert_str_to_num("1.573T"))
print(convert_str_to_num("N/A"))
```

## 1.3) Getting finance data and returns for all tickers (recently listed on IPO)

```
import yfinance as yf
```

```
# Start from an empty dataframe
df_stocks_prices = pd.DataFrame({'A' : []})

# Download all history of stock prices and calculate the future returns for 1-7 days, 30d, 90d, 365d 
# That is: we are very interested if we buy stock at some date (e.g. high EPS) -> if it is going to be a profitable decision
for i,ticker in enumerate(recent_ipos.Ticker):
  yf_ticker = yf.Ticker(ticker)
  historyPrices = yf_ticker.history(period='5Y')
  historyPrices['Ticker'] = ticker

  # Sometimes there is a problem with .index value --> use try
  # https://stackoverflow.com/questions/610883/how-to-know-if-an-object-has-an-attribute-in-python
  try:
    historyPrices['Year']= historyPrices.index.year
    historyPrices['Month'] = historyPrices.index.month
    historyPrices['Weekday'] = historyPrices.index.weekday
    historyPrices['Date'] = historyPrices.index.date
  except AttributeError:
    print(historyPrices.index)
  
  # !!! Important: we do historyPrices['Close'].shift(1) - to get the Close market price the day BEFORE current
  # !!! Important: we do historyPrices['Close'].shift(-i)) - to get the Close market price the i days AFTER current
  #   If you divide second on first -> you get the returns from holding "i days" the stock that you bought the day before financial reporting occurred
  
  for i in [1,2,3,4,5,6,7,30]:
    historyPrices['r'+str(i)] = np.log(historyPrices['Close']/historyPrices['Close'].shift(i) )

  # !!! Important : we generate the cumulative sum of returns for one stock. We can sum it up, as it is np.log for r1
  historyPrices['cumsum_returns'] = historyPrices['r1'].cumsum()

  historyPrices['years_from_now'] = historyPrices['Year'].max()- historyPrices['Year']
  historyPrices['ln_volume']= np.log(historyPrices['Volume'])

  if df_stocks_prices.empty:
     df_stocks_prices = historyPrices
  else: 
    df_stocks_prices = pd.concat([df_stocks_prices,historyPrices], ignore_index=True)
```

```
# 3 new metrics: 
  #  - rise in % in prices last_to_first_trading_day, 
  #  - rise in % for snp500 for an equivalent trading period
  #  - Adjusted_stock_price_change = last_to_first_trading_day - snp500_change
recent_ipos_full["Change_last_price_to_first_trading_day"] = recent_ipos_full.Last_trading_day_close / recent_ipos_full.First_trading_day_close -1
recent_ipos_full["Change_snp500_last_to_first_day"] = recent_ipos_full.snp500_lastday_close / recent_ipos_full.snp500_firstday_close -1
recent_ipos_full["Adj_Change_last_price_to_first"] = recent_ipos_full.Change_last_price_to_first_trading_day - recent_ipos_full.Change_snp500_last_to_first_day
```

```
# Convert S&P500 change in % to an equivalent average daily Log-return metric
recent_ipos_full['snp_avg_return_daily'] = np.log(1+ (recent_ipos_full.Change_snp500_last_to_first_day / recent_ipos_full.r1_count))
```

```
# Trenor ratio = risk-adjusted growth over the benchmark (snp500)
# https://www.investopedia.com/terms/s/sharperatio.asp#what-is-the-sharpe-ratio
# https://www.investopedia.com/terms/t/treynorratio.asp
# Treynor ratio: https://www.investopedia.com/ask/answers/010815/what-difference-between-sharpe-ratio-and-traynor-ratio.asp
recent_ipos_full['Sharpe_Ratio'] = ( (1+recent_ipos_full.r1_mean - recent_ipos_full.snp_avg_return_daily)**365) / (recent_ipos_full.r1_stddev * np.sqrt(365))
```

```
# Returns compared to pre-IPO price
merged_df['total_returns_to_current_date'] = np.log(merged_df['Close'] / merged_df['Initial_Price'])
```

```
# We need to get exact business days Mon-Fri between the IPO date and the current date, rather than just the different in days incl. weekends
# https://stackoverflow.com/questions/3615375/count-number-of-days-between-dates-ignoring-weekends       

# start_array, end_array as datetime
def get_business_days_count(start_array,end_array):
  rez=[]
  for start,end in zip(start_array,end_array):
    daydiff = end.weekday() - start.weekday()
    days = ((end-start).days - daydiff) / 7 * 5 + min(daydiff,5) - (max(end.weekday() - 4, 0) % 5)
    rez.append(days)
  return rez
```

```
# need to generate number of business days when the stock was open
merged_df['business_days_ipo_to_current_trading_day'] = get_business_days_count(pd.to_datetime(merged_df.IPO_Date),pd.to_datetime(merged_df.Date))
```

```
# EXAMPLE FOR ONE STOCK :
def cum_returns_one_ticker(ticker):
  filter = merged_df.Ticker==ticker
  company = merged_df[merged_df.Ticker==ticker].Issuer.max()
# merged_df[filter].head()
  plt.rcParams["figure.figsize"] = (10,6)
  fig,ax = plt.subplots()
  ax.plot( merged_df[filter].business_days_ipo_to_current_trading_day, merged_df[filter].cumsum_returns, 'b-')
  ax.plot( merged_df[filter].set_index("business_days_ipo_to_current_trading_day").rolling(7).cumsum_returns.mean(), 'r--',label='7d-average cumsum r1-returns')
  plt.title(f'Cumulative returns for stock {ticker} -- {company}')
  plt.xlabel('Business days after IPO')
  plt.ylabel('Cumulative return')
  plt.legend(['Cumulative return','7-day average cumulative return'])
  plt.grid(True)
```

## XX.1 Average performance after IPO with confidence intervals
It is too wide to say something -> in average it can be anything

```
# https://jamesrledoux.com/code/group-by-aggregate-pandas
# grouped_single = df.groupby('Team').agg({'Age': ['mean', 'min', 'max']})

# Make one function to show the average returns with Conf. intervals and possibility to select a filter

def show_agg_cumulative_returns(filter):

  agg_info_cumsum_returns = merged_df[filter].groupby(by='business_days_ipo_to_current_trading_day').agg({'cumsum_returns': ['mean', 'min', 'max','std','count']})
  agg_info_cumsum_returns.columns = agg_info_cumsum_returns.columns.droplevel()
  agg_info_cumsum_returns = agg_info_cumsum_returns.reset_index()

  # Conf. intervals
  agg_info_cumsum_returns['ci_up'] =  agg_info_cumsum_returns['mean'] + 1.96 * agg_info_cumsum_returns['std']
  agg_info_cumsum_returns['ci_down'] =  agg_info_cumsum_returns['mean'] - 1.96 * agg_info_cumsum_returns['std'] 

  print(agg_info_cumsum_returns.head())

  # https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html
  ax = agg_info_cumsum_returns[filter].plot(kind='line', 
                                            x='business_days_ipo_to_current_trading_day', 
                                            y='mean',
                                            grid=True)
  # ax.fill_between()
  ax.fill_between(x=agg_info_cumsum_returns[filter]['business_days_ipo_to_current_trading_day'],
                  y1=agg_info_cumsum_returns[filter]['ci_down'],
                  y2=agg_info_cumsum_returns[filter]['ci_up'],
                  alpha=0.2)
```

```
filter = (merged_df.business_days_ipo_to_current_trading_day <= 30) &  (merged_df.business_days_ipo_to_current_trading_day >0)
show_agg_cumulative_returns(filter)
```

```
# https://www.statsmodels.org/dev/examples/notebooks/generated/ols.html
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std

def predict_draw(df):
  X= df[['days_from_ipo_to_now','Initial_Price','Sector_Consumer Discretionary',
        'Sector_Consumer Staples','Sector_Financials','Sector_Healthcare','Sector_Industrials','Sector_Materials',
        'Sector_Real Estate','Exchange_NASDAQ','Realized_offer_amount_m_usd']]
  res = sm.OLS(df["Change_last_price_to_initial"], X).fit()
  print(res.summary())
  print('Parameters: ', res.params)
  print('Standard errors: ', res.bse)
  print('Predicted values: ', res.predict())
  
  x = X['days_from_ipo_to_now']
  y = df["Change_last_price_to_initial"]
  y_true = res.predict()

  prstd, iv_l, iv_u = wls_prediction_std(res)

  fig, ax = plt.subplots(figsize=(8,6))

  ax.plot(x, y, 'o', label="data")
  ax.plot(x, y_true, 'b-', label="True")
  ax.plot(x, res.fittedvalues, 'r--.', label="OLS")
  ax.plot(x, iv_u, 'r--')
  ax.plot(x, iv_l, 'r--')
  ax.legend(loc='best');

  return;
```
