Plot all the library analysis results together for the figure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# sns.set_context('paper')
# sns.set_palette('colorblind')

plt.style.use('aa_paper')
%matplotlib inline
```

```python
bassik = pd.read_csv('../dat/bassik/1kg_wtc_bassik_results.tsv', sep='\t')
bassik['Library'] = 'Bassik'
brunello = pd.read_csv('../dat/brunello/1kg_wtc_brunello_results.tsv', sep='\t')
brunello['Library'] = 'Brunello'
sl = pd.read_csv('../dat/sabatini_lander/1kg_wtc_sabatini_lander_results.tsv', sep='\t')
sl['Library'] = 'Sabatini Lander'
```

```python
df_overall = pd.concat([bassik, brunello, sl])
```

```python
df_overall = df_overall.replace('Common variants\n (MAF >= 5%)\n 1000 Genomes','Common variants (MAF >= 5%)\n 1000 Genomes')
df_overall = df_overall.replace('Sabatini Lander','Sabatini\nLander')
```

```python
plt.figure()
p = sns.catplot('n_vars', kind='count', row='Dataset', hue='Library', 
                data=df_overall, height=1.6, aspect=3.25, sharex=False,
               legend=False, sharey=True, palette='Dark2')
(p.set_axis_labels('Number of Variants', 'Number\n of sgRNAs\n (log scale)')
 .set_titles('{row_name}', verticalalignment='top'))
p.fig.get_axes()[0].set_yscale('log')
p.fig.suptitle('Variants in commonly\nused gRNA libraries')
plt.legend(bbox_to_anchor=(1.05, 0.85),loc=5, frameon=False)
p.fig.set_size_inches(3.0,2.8)

# p.savefig('../figs/grna_libraries_all_wtc_1kgp_countplots.pdf', dpi=300)
```

```python
df_kgp = df_kgp.replace('Sabatini Lander','SL')
fig = plt.figure(figsize=(2.1,2.1))
ax1 = fig.add_axes([0.1, 0.1, 1.0,1.0])
p = sns.countplot(x='n_vars', hue='Library', data=df_kgp, palette='Dark2', ax=ax1)
plt.yscale('log', ax=ax1)
plt.xlabel('Number of variants in sgRNA')
plt.ylabel('Number of sgRNAs (log scale)')
plt.title('Common variants in sgRNAs\n in widely used libraries')
ax1.get_legend().remove()

# add "histogram legend"
perc_df = pd.DataFrame()

libraries = df_kgp.Library.unique()
percs = []

for library in libraries:
    percs.append(100.0 * len(df_kgp.query('(n_vars > 0) and (Library == @library)')) / len(df_kgp.query('Library == @library')))
    
perc_df['Library'] = libraries
perc_df['% sgRNAs'] = percs
ax2 = fig.add_axes([0.75, 0.78, 0.3, 0.3], frameon=False)
plt.xticks(range(0, 6, 1), size=11)
plt.yticks(size=11)
q = sns.barplot(y='Library', x='% sgRNAs', data=perc_df, palette='Dark2',
               ax=ax2)

# # save to file
# fig.savefig('../figs/grna_libraries_1kgp_only_n_sgrnas.pdf', dpi=300,
#                       bbox_inches='tight')
```
