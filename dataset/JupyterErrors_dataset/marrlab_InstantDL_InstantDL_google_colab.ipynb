Import the deep learning pipeline from InstantDL

```
from instantdl import GetPipeLine
```

Please specify the configs and run InstantDL with the cell below

```
configs = { "use_algorithm" : "Classification",
           "path" : "path",
           "pretrained_weights" : None, 
            "batchsize" : 10, 
            "iterations_over_dataset" : 50, 
            "data_gen_args" : false, 
            "loss_function" : "mse", 
            "num_classes" : 10, 
            "image_size" : false,
            "calculate_uncertainty" : false,
            "evaluation" : true}

P = GetPipeLine(configs=configs)
P.run()
```
