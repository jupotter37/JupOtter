## Array do NumPy

vetor unidimensional. Shape e type são importantes para caracterizar o array.

```python
import numpy as np
a = np.array([1,2,3,4])
print(a[::2])
print(a.shape)
print(a.dtype)
```

```python
a = np.array([[1,2,3,4],
              [5,6,7,8]])
print(a.shape, a.dtype)
print(a)
print(a[1,2])
n,m = a.shape
print ('a.ravel', a.ravel())
print(a[0:n,0:m])
print(a[:,:])
```

```python
print(a)
print('só as 2 primeiras colunas')
print(a[:,0:1])
print ('só a última coluna, mas continuando como matriz')
print(a[:,-1:])
print(a[1,:].shape)
print(a[1,:])
```

```python
print(a)
print(a[::-1,::-1])
```
