# Modularizing RF

```python
df = pd.read_csv("./CSVs/AXP_quarterly_financial_data.csv", delimiter = ',')
df.head()
```

```python
'''

Function that will take a Dataframe as an Input
Removes all the unwanted columns
And then Trains Random forest models to Predict 
next quarter price change

input : Dataframe of Stock Fundamental Data
        41 Columns
        
output : Predictions for last 20% of the Quarters
         Actual Values for last 20% of the Quarters

'''
def get_empty_columns_names(df):
    s1 = df.isna().sum()
    vals = s1.values
    indexes = s1.index
    
    threshold = df.shape[0]/2
    
    return_list =[]
    
    for i in range(len(s1)):
        if s1[i]>threshold:
            return_list.append(indexes[i])
    
    return return_list

from sklearn.model_selection import train_test_split
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

def Get_Random_Forest_Predictions(df):
    
    #MAKING QUARTER END DATETIME FORMAT
    df['Quarter end'] = pd.to_datetime(df['Quarter end'])
    
    #Sorting the dates to get time series data
    df = df.sort_values(by=['Quarter end'])
    
    '''

    This is done to replace all 'None' Values in the dataframe with 
    the previous valid value

    '''
    for column in df.columns:
        df[column] = df[column].replace('None',value= np.NaN)
    
    #DROPPING UNWANTED COLUMNS
    df = df.drop(columns = ['Quarter end','Price high', 'Price low','Shares', 'Shares split adjusted', 'Split factor','Non-controlling interest',
                            'Preferred equity','Goodwill & intangibles','Long-term debt','Dividend per share',
                            'Dividend payout ratio','Long-term debt to equity ratio','Free cash flow per share','Capital expenditures',
                       ])
    #GETTING THE PRICE AS A SEPARATE COLUMN
    y_prev = df['Price']
    df = df.drop(columns = ['Price'])
    
    new_df = df
    names = get_empty_columns_names(new_df)
    new_df = new_df.drop(columns=names)
    new_df = new_df.fillna(method='bfill')
    
    
    for column in df.columns:
        new_df[column] = new_df[column].astype('float64')
    
    #GETTING PERCENTAGE CHANGE VALUES
    new_df = new_df.pct_change()
    
    
    '''
    
    Removing the first row of NaN values
    for both X and Y

    '''

    new_df = new_df[1:]


    X = new_df


    Y = y_prev.pct_change()

    Y = Y[1:]
    
    X = pd.concat([X,Y], axis = 1)
    
    #Getting Next Quarter %change of price 
    #This will be our target variable

    Target = Y[1:]
    X = X[:-1]
    
    Y = Target
    X = X.values
    Y = Y.values
    X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
    
    X_train_means = np.mean(X_train, axis = 0)
    X_train_std = np.mean(X_train, axis = 0)
    
    X_train_scaled = (X_train - X_train_means)/X_train_std

    X_test_scaled = (X_test - X_train_means)/X_train_std
    # Instantiate model with 1000 decision trees
    rf = RandomForestRegressor(n_estimators = 1000,max_depth = 30 ,max_features = 'sqrt',min_samples_leaf=4, random_state = 42)
    # Train the model on training data
    rf.fit(X_train, y_train)
    
    # Use the forest's predict method on the test data
    predictions = rf.predict(X_test)
    # Calculate the absolute errors
    errors = abs(predictions - y_test)
    # Print out the mean absolute error (mae)
    print('Mean Absolute Error:', round(np.mean(errors), 2))
    # Calculate mean absolute percentage error (MAPE)
    mape = 100 * (errors / y_test)
    # Calculate and display accuracy
    accuracy = 100 - np.mean(mape)
    print('Accuracy:', round(accuracy, 2), '%.')
    
    
    return predictions,y_test
```
