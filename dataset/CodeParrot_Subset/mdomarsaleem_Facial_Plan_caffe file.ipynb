```python
import os,sys,re
import numpy as np
os.chdir('/home/mckc/pycaffe/')
direct = os.getcwd()
```

```python
files = os.listdir(os.getcwd())
subject = []
file_name = []
data_type = []
```

```python
for name in files:
    file_name = np.append(file_name,direct+'/' + name)
    print name
    index = name.index('_')
    data_type = np.append(data_type,name[index:index+4])
    subject = np.append(subject,name[:index])
```

```python
map, Y_number = np.unique(subject, return_inverse=True)
```

```python
Y_number
```

```python
data = np.vstack((np.array(file_name), np.array(Y_number),np.array(data_type)))
data = np.transpose(data)
data = data.astype(str)
```

```python
train = data[data[:,2]=='_tra',]
test = data[data[:,2]!='_tra',]
```

```python
np.savetxt("train.txt", train[:,:2], delimiter=" ",fmt='%s')
np.savetxt("test.txt", test[:,:2], delimiter=" ",fmt='%s')
```
