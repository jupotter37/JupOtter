# 測試完畢，統整寫for迴圈

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
import os


total_ans = []

for page in range(1,5):
    
    #提取資料
    page_data = list_train_order_value[page]
    
    page_data = np.trim_zeros(page_data, trim='f')  #拿掉前面有太多0的，會有bias
    
    page_data_label = page_data[10:] #每10組給出一次答案，所以答案不包含前10組
    
    train_feature = []
    for i in range(len(page_data_label)):
        train_feature.append(page_data[i:i+10])
    train_feature_array = np.array(train_feature)
    
    
    #標準化
    train_feature_array = train_feature_array/max(page_data)
    page_data_label = page_data_label/max(page_data)

    #print(train_feature_array) #測試用
    #print(page_data_label) #測試用
    
    
    #訓練模型
    model = Sequential() 
    
    model.add(Dense(units=100, 
                    input_dim=10, 
                    kernel_initializer='uniform'))
    model.add(Dropout(0.5))

    model.add(Dense(units=100,  
                    kernel_initializer='uniform'))
    model.add(Dropout(0.5))

    model.add(Dense(units=1))

    #print(model.summary())  #測試用

    '''
    weights_path = 'Savemodels/WIKI(Kaggles)_MLP.h5'

    if os.path.isfile(weights_path): 
        model.load_weights(weights_path) #只要跑過一次就取出來直接使用
    else:
        pass
    '''
    
    model.compile(loss='mse',   
                  optimizer='adam', metrics=['accuracy'])

    model.fit(x=train_feature_array, y=page_data_label,  
              validation_split=0.2, epochs=20, batch_size=200, verbose=1)
    
    #model.save_weights("Savemodels/WIKI(Kaggles)_MLP.h5") #方便下一次跑?
    
    
    
    #訓練後跑答案
    ans_list = train_feature_array[-1].reshape(-1,10)
    rand_list = np.random.randn(60) #常態分佈的雜訊
    
    for i in range(60): #做好做滿60次預測
        ans = model.predict(ans_list[i:i+10].reshape(-1,10))
        ans_add_noise = ans + 0.2*np.std(train_feature_array)*rand_list[i]
        #避免ans太快收斂，加入rand條件讓他上下波動
        
        if ans_add_noise < 0:
            ans_add_noise = 0  #瀏覽人次沒有在<0的啦～
        else:
            ans_add_noise = ans_add_noise
        
        ans_list = np.append(ans_list, ans_add_noise) 
                                            
    
    ans_list_round = np.round(ans_list*np.max(page_data)) #四捨五入成整數的瀏覽人次
    real_ans = ans_list_round[11:] #最終每組60個數字的答案
    
    
    total_ans.append(real_ans) #把每一組的60個數字通通加總在一張表上，最後再貼回去答案卷
    print('case completed: ', page)
    
    
print('All cases completed!!')
```
