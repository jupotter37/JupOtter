```python
# Example of Linear Regression with Keras and TensorFlow
# Y = 10 * X +100
```

```python
from __future__ import absolute_import, division, print_function

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(1,)),
    keras.layers.Dense(50, activation=tf.nn.relu),
    keras.layers.Dense(50, activation=tf.nn.relu),
    keras.layers.Dense(50, activation=tf.nn.relu),
    keras.layers.Dense(1)    
])
```

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(0.001), 
              loss='mean_squared_error',
              metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
import random

N = 20000
arrayy = np.arange(0, N) 
input_values = arrayy[np.indices((N, 1))[0]]

output_values_0 = np.multiply(input_values, 10)
# add some noise (deliberately)
output_values_1 = np.add(output_values_0, random.randint(0,1))
output_values = np.add(output_values_1, 100)
```

```python
model.fit(input_values, output_values, epochs=1000)
```
