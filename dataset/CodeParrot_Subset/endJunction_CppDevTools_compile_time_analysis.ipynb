```python
import json
import matplotlib
%matplotlib inline
```

```python
f = open('timing.json')
data = json.load(f)
f.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
```

```python
df.plot(y='time')
```

```python
df.sort('time').tail(40).plot(x='file', y='time', kind='bar', figsize=(16, 6), title='Compilation time in seconds')
```
