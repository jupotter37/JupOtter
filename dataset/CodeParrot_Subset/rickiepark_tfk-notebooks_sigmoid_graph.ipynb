```python
x = np.arange(-10., 10., 0.1)
y = 1 / (1 + np.exp(-x))
```

```python
plt.ylabel('Y')
plt.ylim([0, 1.006])
plt.xlabel('X')
plt.plot(x, y)
plt.axhline(0.5, color='red')
plt.axvline(0, color='red')
plt.show()
```

```python
dy = y * (1 - y)
```

```python
plt.ylabel('dy')
plt.ylim([0, 0.27])
plt.xlabel('X')
plt.plot(x, dy)
plt.axvline(0, color='red')
plt.show()
```
