```python
import yt
```

```python
ds = yt.load("gen-mesh-one-material_out.e", step=-1)

ad = ds.all_data()

temp = ad[('connect1', 'temp')]
group1 = ad['group1']
group2 = ad['group2']

print(temp.max())
print(temp.min())
print(group1.max())
print(group1.min())
print(group2.max())
print(group2.min())

g1slice = yt.SlicePlot(ds, 'z', 'group1')
g2slice = yt.SlicePlot(ds, 'z', 'group2')
```

```python
g1slice.set_log('group1', False)
g1slice.set_zlim('group1', group1.min(), group1.max())
```

```python
g2slice.set_log('group2', False)
g2slice.set_zlim('group2', group2.min(), group2.max())
```
