```python
# imports
import numpy as np

from matplotlib import pyplot as plt

from dla_cnn import io as dla_io
```

## Load

```python
g16_abs = dla_io.load_garnett16()
```

## Cutting on pDLA

### Histogram pDLAD

```python
gdp = g16_abs['pDLAD'] > 0.
```

```python
pDLAD = g16_abs['pDLAD'][gdp]
```

```python
plt.clf()
ax = plt.gca()
ax.hist(pDLAD)
# Label
ax.set_xlabel('pDLAD')
plt.show()
```

### Cumulative

```python
srt = np.argsort(pDLAD)
```

```python
ydum = np.arange(len(pDLAD)) / len(pDLAD)
```

```python
plt.clf()
ax = plt.gca()
ax.plot(pDLAD[srt], ydum)
#
ax.set_ylabel('Cumulative')
ax.set_xlabel('pDLAD')
plt.show()
```

### Cut at 0.95?

```python
for pcut in [0.9, 0.95, 0.99, 0.999]:
    print(pcut, np.sum(pDLAD > pcut))
```
