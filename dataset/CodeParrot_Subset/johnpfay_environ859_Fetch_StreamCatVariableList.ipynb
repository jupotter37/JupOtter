## List stream cat variables

```python
import pandas as pd
```

```python
theURL = 'ftp://newftp.epa.gov/EPADataCommons/ORD/NHDPlusLandscapeAttributes/StreamCat/Documentation/VariableList-QuickReference.html'
```

```python
dfList = pd.read_html(theURL,skiprows=6)
dfRaw = dfList[0]
```

```python
dfRaw.head()
```

```python
dfRaw.columns = ['Index','Variable','Description','Table']
dfClean = dfRaw.drop('Index',axis=1)
dfClean.dropna(inplace=True)
dfClean.head()
```

```python
#List the variables includeed in the NLCD2011_<StateID>.csv
dfClean[dfClean['Table'].str.contains("NLCD2006")]  #['Variable']
```
