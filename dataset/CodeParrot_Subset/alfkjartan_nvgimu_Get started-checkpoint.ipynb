## The content of the raw IMU file
The columns of the IMU data contain: 
0: Packet number, 
1: Gyroscope X (deg/s),
2: Gyroscope Y (deg/s),
3: Gyroscope Z (deg/s),
4: Accelerometer X (g),
5: Accelerometer Y (g),
6: Accelerometer Z (g),
7: Magnetometer X (G),
8: Magnetometer Y (G),
9: Magnetometer Z (G)
