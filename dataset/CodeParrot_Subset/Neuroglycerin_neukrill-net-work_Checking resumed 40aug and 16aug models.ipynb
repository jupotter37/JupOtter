```python
def make_curves(model, *args):
    curves = None
    for c in args:
        channel = m.monitor.channels[c]
        c = c[0].upper() + c[1:]
        if not curves:
            curves = hl.Curve(zip(channel.example_record,channel.val_record),group=c)
        else:
            curves += hl.Curve(zip(channel.example_record,channel.val_record),group=c)
    return curves
```

```python
means = [c for c in sorted(m.monitor.channels.keys()) if "mean" in c and "norm" in c]
make_curves(m,*means)
```
