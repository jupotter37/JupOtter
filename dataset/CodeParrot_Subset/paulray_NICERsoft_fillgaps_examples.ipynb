```python
import numpy as np
import matplotlib.pyplot as plt
from nicer.fillgaps import fillgaps
```

# NICER Pulsars
### PSR_J0218+4232

```python
# MJDs/residuals for PSR_J0218+4232
mjds = []
residuals = []
with open("0218-grid-nicer.txt", 'r') as file:
    line = file.readline()
    while line:
        vals = line.split()
        if vals != [] and vals[0][0] != '#':
            mjds.append(int(float(vals[0])))
            residuals.append(float(vals[1]))
        line = file.readline()
```

```python
x = np.array(mjds)
xindices = np.array((x - x[0]), dtype=int)
ydense = np.zeros(xindices.max() + 1) * np.nan
ydense[xindices] = residuals
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("MJD (day)")
ax.set_ylabel("Residual (us)")
ax.plot(ydense)
```

```python
fillgaps(ydense, "reflect+invert")
fig, ax = plt.subplots()
ax.plot(ydense)
```

# NANOGrav Pulsars
### J1747-4036

```python
resid1, resid2, resid3, resid4 = [], [], [], []
resids = [resid1, resid2, resid3, resid4]
i = 0
with open("grid.out", 'r') as file:
    line = file.readline()
    while line:
        vals = line.split()
        if vals != [] and vals[0][0] != '#':
            resids[int(i)].append(float(vals[1]))
        if vals == []:
            i = i+0.5 # 2 blank lines
        line = file.readline()

fig, ax = plt.subplots()
ax.set_xlabel("MJD (30-day bin)")
ax.set_ylabel("Residual (us, 30-day average)")
ax.plot(resids[0])
```

```python
fillgaps(resids[0], "reflect+invert")
fig, ax = plt.subplots()
ax.plot(resids[0])
```
