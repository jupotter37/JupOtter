```python
year = 2014
prefCode = 11
regionCode = 1
countryCode = 103
def get_foreigners(key, year, prefCode):
    x = json.loads(requests.get('https://opendata.resas-portal.go.jp/api/v1/tourism/foreigners/forFrom?' + \
                                'purpose=1&year=%d&prefCode=%d' % (year, prefCode), headers={'X-API-KEY':key}).text)
    print x
    return x['result']
pref = get_foreigners(key, year, prefCode)
```

```python
p2c = {}
c2p = {}
for year in range(2015, 2017, 1):
    for pC in range (1, 48, 1):
        res = get_foreigners(key, year, pC)
        if pC not in p2c:
            p2c[pC] = Counter()
        tmp = res['changes']
        for i in xrange(len(tmp)):
            cC = int(tmp[i]['countryCode'])
            if cC not in c2p:
                c2p[cC] = Counter()
            data = tmp[i]['data']
            for j in xrange(len(data)):
                val = data[j]['value']
                c2p[cC][pC] += val
                p2c[pC][cC] += val
```

```python
# list化、保存
lists = []
for pC in p2c:
    for cC in p2c[pC]:
        lists.append([pC, cC, p2c[pC][cC]])
df = pd.DataFrame(np.asarray(lists), columns=None)
df.to_csv('data/list_pc.csv', index=False)
```
