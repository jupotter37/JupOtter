```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#Read CO2 data
CO2 = pd.read_csv('CO2.csv', header = 0, 
                  names = ['Date_Time','CO2','TimeFix'])

#plot the CO2 feature
CO2.plot.line('Date_Time','CO2',figsize=(20,5))
plt.show()
```

```python
#Read bluetooth data 
Bluetooth = pd.read_csv('Bluetooth.csv', delimiter = ";", header = 0, 
                        names = ['Date_Time','Location','Bluetooth','TimeFix'])

#plot the Bluetooth feature
Bluetooth.plot.line('Date_Time','Bluetooth',figsize=(20,5))
plt.show()
```

```python
#Read light date
Light = pd.read_csv('Light.csv', delimiter = ";", header = 0, names = ['Date_Time','Location','Light','TimeFix'])

#plot Light feature
Light.plot.line('Date_Time','Light',figsize=(20,5))
plt.show()
```

```python
#Read sound date
Sound = pd.read_csv('Sound.csv', header = 0, 
                    names = ['Date_Time','Sound','TimeFix'])

#Plot sound feature
Sound.plot.line('Date_Time','Sound',figsize=(20,5))
plt.show()
```

```python
#Read temperature and humidity data
Temp_humidity = pd.read_csv('Temp_humidity.csv', delimiter = ";", header = 0, 
                            names = ['Date_Time','Location','Temp','Humidity','Time fix'])

#Plot temperature data
Temp_humidity.plot.line('Date_Time','Temp',figsize=(20,5))
plt.show()

#Plot humidity data
Temp_humidity.plot.line('Date_Time','Humidity',figsize=(20,5))
plt.show()
```
