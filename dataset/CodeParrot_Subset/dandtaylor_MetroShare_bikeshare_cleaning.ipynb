# Get the bikeshare data ready for analysis

```python
import pandas as pd
import pickle
```

#### import 2015-Q4 through 2016-Q3 bike data

```python
bikes5 = pd.read_csv('2015-Q4-Trips-History-Data.csv')
bikes6 = pd.read_csv('2016-Q1-Trips-History-Data.csv')
bikes7 = pd.read_csv('2016-Q2-Trips-History-Data.csv')
bikes8 = pd.read_csv('2016-Q3-Trips-History-Data-1.csv')
bikes9 = pd.read_csv('2016-Q3-Trips-History-Data-2.csv')
bikes7.rename(columns={'Account type':'Member Type'}, inplace=True)
bikes5.rename(columns={'Bike #':'Bike number', 'Member type':'Member Type'}, inplace=True)

bikes = pd.concat([bikes5, bikes6, bikes7, bikes8,
                    bikes9], ignore_index=True)
```

```python
bikes.head()
```

Create DatetimeIndex from `Start date`

```python
bikes['Start date'] = pd.to_datetime(bikes['Start date'], format='%m/%d/%Y %H:%M')
```

```python
bikes = bikes.set_index('Start date')
```

save this DataFrame with pickle

```python
pickle.dump( bikes, open( "bikeshare_rides_all.p", "wb" ) )
```
